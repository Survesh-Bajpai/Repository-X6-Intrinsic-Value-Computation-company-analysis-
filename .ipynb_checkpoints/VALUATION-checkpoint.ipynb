````python
# %% [markdown]
# # Dynamic Stock Financial Analysis & DCF Valuation
# **(Interactive Report - Dynamic/User Input Driven Assumptions - V9 - Revenue CAGR Initial Growth)**
#
# **Analyst:** Simulated AI Research Analyst
#
# **Objective:** To perform a comprehensive financial analysis and estimate the intrinsic value per share of a publicly traded company using the Discounted Cash Flow (DCF) methodology. Prioritizes dynamic fetching, falls back to user input, and uses configured defaults as a last resort. Includes fallback benchmarks for ratios and robust error handling/display. Adapts currency formatting for Indian stocks. **Initial growth uses last 3 years of Revenue CAGR with bounds and default.**
#
# **Methodology:**
# 1.  **User Input & Dynamic Fetching:** Get ticker. Attempt dynamic Rf, require user confirmation/input for Rf, Rm, Beta (if needed), Credit Spread (if needed), and Ratio Benchmarks (with fallbacks). Detects Indian stocks for currency formatting.
# 2.  **Data Acquisition:** Use `yfinance` for statements, info, market data. Fetch extended price history.
# 3.  **WACC Calculation:** Determine WACC using fetched/user inputs via CAPM. Handle errors.
# 4.  **FCFF Calculation & Projection:** Calculate historical FCFF. **Estimate initial growth using 3-year Revenue CAGR, apply 1.1%-15% bounds, default to 6.5% if data unavailable.** Project FCFF forward using linear decay to terminal growth. Handle errors.
# 5.  **Terminal Value:** Estimate value beyond the explicit forecast period.
# 6.  **Discounting & Valuation:** Discount cash flows and terminal value. Handle errors.
# 7.  **Ratio Analysis:** Calculate ratios and compare to user-input or fallback default benchmarks. Handles non-finite values.
# 8.  **Interactive Visualization:** Present findings using Plotly, including dynamic gauge ranges, robust display logic, and ₹/Lakh/Crore formatting for Indian stocks.

# %% [markdown]
# ## 1. Setup and Libraries

# %%
# Standard Libraries
import warnings
import sys
import numpy as np
import pandas as pd
import datetime
import math # For isnan, isinf checks

# Data Fetching
import yfinance as yf

# Visualization (Using Plotly)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

# Display Utilities
from IPython.display import display, Markdown, HTML

# Settings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pio.templates.default = "plotly_white" # Cleaner Plotly template

# %% [markdown]
# ## 2. Configuration and Helper Functions (Updated Default Initial Growth and Bounds)

# %%
# --- Model Configuration ---
DEFAULT_TICKER = "WIPRO.NS" # Default if user input fails
FORECAST_YEARS = 5
# Terminal Growth Rate CAP (Also acts as a default ceiling)
TERMINAL_GROWTH_RATE_CAP = 0.05 # Default/Cap set to 5%
# ----> NEW: Default Initial Growth Rate (Used if 3yr Rev CAGR fails) <----
DEFAULT_INITIAL_GROWTH_RATE = 0.065  # Default set to 6.5%
# ----> NEW: Initial Growth Rate Bounds <----
MIN_INITIAL_GROWTH_RATE = 0.011 # Lower ceiling 1.1%
MAX_INITIAL_GROWTH_RATE = 0.15  # Upper ceiling 15%

# Default Risk-Free Rate suggestion (used if dynamic fetch fails)
DEFAULT_RF_RATE_SUGGESTION = 6.0 # Default set to 6.0% (as percent)

DEFAULT_CURRENCY_SYMBOL = "₹" # Changed default to Rupee, will be overridden by yfinance if possible
DEFAULT_TAX_RATE = 0.25     # General fallback tax rate if calculation fails
DEFAULT_CREDIT_SPREAD = 0.03 # Fallback spread over Rf for Cost of Debt
DEFAULT_MARKET_RISK_PREMIUM = 0.08 # Used only to suggest Rm to user

# --- Generic Fallback Benchmark Ratios (Used ONLY if User leaves input blank) ---
DEFAULT_BENCHMARK_RATIOS = {
    'P/E': 20.0,
    'P/B': 2.5,
    'EV/EBITDA': 12.0,
    'ROE': 15.0, # As percent value (e.g., 15 for 15%)
    'ROA': 8.0,  # As percent value
    'ROCE': 12.0,# As percent value
    'Asset Turnover': 0.8,
    'Current Ratio': 1.5,
    'Quick Ratio': 1.0
}

# --- Model Stability Bounds --- (Not primary financial assumptions, but guardrails)
MIN_WACC = 0.06
MAX_WACC = 0.25
MIN_TAX_RATE = 0.10
MAX_TAX_RATE = 0.40
MIN_GROWTH_FLOOR = 0.00 # Minimum growth floor (for *terminal* growth primarily now)
# MAX_INITIAL_GROWTH is now defined above (0.15)
MIN_COST_OF_DEBT = 0.03
MAX_COST_OF_DEBT = 0.15

# --- Ratio Keys Needed for Benchmarking ---
REQUIRED_BENCHMARK_RATIOS = list(DEFAULT_BENCHMARK_RATIOS.keys()) # Use keys from default dict

# --- Helper Functions ---
def print_keys(statement, name):
    """Prints available keys in a yfinance statement for debugging"""
    print(f"\n--- Available Keys in {name} ---")
    if isinstance(statement, pd.DataFrame):
        valid_indices = [idx for idx in statement.index if idx is not None]
        print(", ".join(map(str, valid_indices)))
    elif isinstance(statement, dict):
        print(", ".join(statement.keys()))
    print("------------------------------------\n")

def safe_get(data, key_list, statement_name="Statement", default=0.0):
    """
    Safely retrieves a numeric value from yfinance data (Series or Dict).
    Tries multiple keys, returns default if not found/invalid. Handles NaN/inf better.
    """
    if not isinstance(key_list, list):
        key_list = [key_list]

    value = None
    for key in key_list:
      current_value = None
      raw_val = None
      if isinstance(data, pd.Series):
          if key in data.index:
              try:
                  raw_val = data.loc[key]
                  # Check if raw_val is already numeric and finite before converting
                  if isinstance(raw_val, (int, float)) and np.isfinite(raw_val):
                      current_value = float(raw_val)
                  # Handle Series case (sometimes yfinance returns single value Series)
                  elif isinstance(raw_val, pd.Series):
                       if not raw_val.empty and pd.notna(raw_val.iloc[0]) and np.isfinite(raw_val.iloc[0]):
                            current_value = float(raw_val.iloc[0])
                  # Handle string conversion only if necessary and check finiteness
                  elif pd.notna(raw_val):
                      try_val = float(raw_val)
                      if np.isfinite(try_val):
                          current_value = try_val
              except (KeyError, TypeError, IndexError, ValueError):
                   pass # Ignore errors for this specific key/conversion
      elif isinstance(data, dict):
          raw_val = data.get(key)
          if raw_val is not None:
              try:
                  if isinstance(raw_val, str) and raw_val.lower() in ['n/a', 'none', '']:
                      pass # Treat as invalid
                  else:
                      temp_val = float(raw_val)
                      if np.isfinite(temp_val): # Check for NaN/inf after conversion
                           current_value = temp_val
              except (TypeError, ValueError):
                   pass

      if current_value is not None:
            value = current_value
            break # Found a valid value

    if value is None:
        value = default # Use the provided default if no valid value found
    return value

def format_currency(value, symbol="₹", precision=2, is_indian=False):
    """
    Formats number as currency string, handles None/NaN/inf.
    Uses Lakh/Crore for Indian stocks if is_indian=True.
    """
    if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
    if abs(value) > 1e18: return "Inf" # Handle extremely large numbers as Inf

    sign = "-" if value < 0 else ""
    abs_val = abs(value)

    if is_indian:
        # Use ₹ symbol regardless of fetched symbol for consistency
        symbol = "₹"
        if abs_val >= 1e7:  # 1 Crore (10 million)
            formatted_val = f"{abs_val / 1e7:.{precision}f} Cr"
        elif abs_val >= 1e5:  # 1 Lakh (100 thousand)
            formatted_val = f"{abs_val / 1e5:.{precision}f} Lakh"
        else:
            formatted_val = f"{abs_val:,.{precision}f}" # Add comma separators for smaller Rupee values
        return f"{sign}{symbol} {formatted_val}" # Add space after symbol
    else:
        # Standard international formatting (K, M, B, T)
        if abs_val >= 1e12: formatted_val = f"{abs_val / 1e12:.{precision}f}T"
        elif abs_val >= 1e9: formatted_val = f"{abs_val / 1e9:.{precision}f}B"
        elif abs_val >= 1e6: formatted_val = f"{abs_val / 1e6:.{precision}f}M"
        elif abs_val >= 1e3: formatted_val = f"{abs_val / 1e3:.{precision}f}K"
        else: formatted_val = f"{abs_val:.{precision}f}"
        return f"{sign}{symbol}{formatted_val}" # Standard symbol prefix

def format_percent(value, precision=1):
     """Formats number as percentage string, handles None/NaN/inf."""
     if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
     if abs(value) > 1e5: return "Inf" # Handle extreme % values

     return f"{value * 100:.{precision}f}%"

def get_validated_float_input(prompt_message, default_value=None, allow_blank=False, min_val=None, max_val=None):
    """Gets and validates user input for a float value, handles defaults, blanks, and bounds."""
    while True:
        prompt_with_default = prompt_message
        if default_value is not None:
            prompt_with_default += f" [Default: {default_value}]: "
        elif allow_blank:
             prompt_with_default += f" [Leave blank to skip/use fallback]: "
        else:
             prompt_with_default += ": "

        try:
            user_input_str = input(prompt_with_default).strip()

            if not user_input_str:
                if allow_blank: return None
                if default_value is not None:
                    try: return float(default_value)
                    except ValueError: raise ValueError(f"Internal Error: Invalid default value '{default_value}'.")
                raise ValueError("Input cannot be empty.")

            if user_input_str.endswith('%'): # Allow user to add % sign optionally
                user_input_str = user_input_str[:-1]

            value = float(user_input_str)

            if min_val is not None and value < min_val:
                 raise ValueError(f"Value must be >= {min_val}.")
            if max_val is not None and value > max_val:
                 raise ValueError(f"Value must be <= {max_val}.")

            return value
        except ValueError as e:
            print(f"❌ Invalid input: {e}. Please enter a valid number.")

def try_get_dynamic_rf(rf_ticker_symbol="^TNX"):
    """Attempts to fetch a dynamic risk-free rate (e.g., US 10Y Treasury)."""
    print(f"Attempting to fetch dynamic Risk-Free Rate ({rf_ticker_symbol})...")
    try:
        rf_ticker = yf.Ticker(rf_ticker_symbol)
        hist = rf_ticker.history(period="5d")
        if not hist.empty and 'Close' in hist.columns:
            last_close = hist['Close'].iloc[-1]
            if pd.notna(last_close):
                dynamic_rf = last_close / 100.0 # Assume Yahoo gives yield as %
                if 0.001 < dynamic_rf < 0.20: # Check bounds 0.1% to 20%
                    print(f"✅ Dynamically fetched Rf ({rf_ticker_symbol}): {format_percent(dynamic_rf)}")
                    return dynamic_rf
                else:
                     print(f"⚠️ Fetched Rf value ({format_percent(dynamic_rf)}) out of typical range. Ignoring.")
                     return None
            else:
                 print(f"⚠️ Fetched Rf data was invalid (NaN) for {rf_ticker_symbol}.")
                 return None
        else:
             print(f"⚠️ Could not fetch recent data for {rf_ticker_symbol}.")
             return None
    except Exception as e:
        print(f"❌ Error fetching dynamic Rf ({rf_ticker_symbol}): {e}")
        return None

# --- Growth Calculation Helpers ---
def calculate_average_growth(values_list):
    """
    Calculates average growth rate from a list of historical values (newest first).
    Handles NaNs, skips growth from zero/negative bases. Returns NaN if no valid growth periods.
    """
    growth_rates = []
    if len(values_list) < 2:
        return np.nan # Need at least two points for one growth rate

    for i in range(len(values_list) - 1):
        val_t = values_list[i]       # More recent value (e.g., T-0)
        val_t_minus_1 = values_list[i+1] # Older value (e.g., T-1)

        if pd.notna(val_t) and pd.notna(val_t_minus_1) and math.isfinite(val_t) and math.isfinite(val_t_minus_1):
            # Check if base year is positive and non-negligible
            if val_t_minus_1 > 1e-6: # Use a small threshold instead of exact zero
                g = (val_t / val_t_minus_1) - 1
                growth_rates.append(g)
            # else: # Optional: log skipped period
            #    print(f"   - Skipping growth calc period {i+1}-{i+2}: Invalid base value {val_t_minus_1}")
        # else: # Optional: log skipped period
        #    print(f"   - Skipping growth calc period {i+1}-{i+2}: NaN/inf values ({val_t}, {val_t_minus_1})")


    if not growth_rates:
        return np.nan
    else:
        return np.mean(growth_rates)

def calculate_cagr(start_val, end_val, num_years):
    """Calculates Compound Annual Growth Rate. Requires positive start_val."""
    if pd.isna(start_val) or pd.isna(end_val) or not math.isfinite(start_val) or not math.isfinite(end_val) or num_years <= 0:
        return np.nan
    if start_val <= 1e-9: # Check for non-positive start value
        return np.nan

    # Store original sign of end value before adjustments
    original_end_val_sign = np.sign(end_val)

    # Adjust end_val if non-positive to allow calculation
    if end_val <= 1e-9:
        end_val = 1e-9 # Use tiny positive for calculation

    try:
        # Calculate CAGR based on absolute ratio
        # Handle potential complex number if end_val is negative and num_years causes fractional exponent
        # by calculating based on absolute value first.
        ratio = abs(end_val / start_val)
        if ratio < 0: # Should not happen with abs(), but as safeguard
             return np.nan

        cagr = (ratio) ** (1 / num_years) - 1

        # If the original end value was negative, reflect this as a negative CAGR magnitude
        # A true CAGR isn't really defined for negative end values, but this shows the direction.
        if original_end_val_sign < 0:
             # Make it significantly negative to indicate endpoint negativity
             # This indicates a collapse rather than negative growth in the usual sense
             cagr = -(abs(cagr) + 1) * 1.5 # e.g. a -5% CAGR becomes -(1.05)*1.5 = -1.575 (-157.5%)

    except (ValueError, OverflowError, ZeroDivisionError):
        cagr = np.nan
    return cagr
# --- End Growth Calculation Helpers ---

# %% [markdown]
# ## 3. User Input & Dynamic Fetching: Ticker, Rf, Rm, Benchmarks, Beta, Spread

# %%
input_sources = {}

# --- Get Ticker Symbol ---
user_ticker = input(f"Enter stock ticker (e.g., MSFT, AAPL, RELIANCE.NS) [Default: {DEFAULT_TICKER}]: ")
if not user_ticker: user_ticker = DEFAULT_TICKER
ticker_symbol = user_ticker.strip().upper()
display(Markdown(f"**Analyzing Ticker:** `{ticker_symbol}`"))

# ----> DETECT IF INDIAN STOCK for currency formatting <----
is_indian_stock = ticker_symbol.endswith(".NS")
if is_indian_stock:
    print("Info: Detected Indian stock (.NS). Using ₹/Lakh/Crore formatting.")

# --- Determine RF Ticker based on Stock Ticker ---
rf_proxy_ticker = "^TNX" # Default US 10Y
if is_indian_stock: rf_proxy_ticker = "^NS10"; print(f"Info: Using Indian Rf proxy: {rf_proxy_ticker}")
elif ticker_symbol.endswith(".L"): rf_proxy_ticker = "^FTLC"; print(f"Info: Using UK Rf proxy: {rf_proxy_ticker}")
# Add more country/exchange specific Rf proxies if needed

# --- Get Risk-Free Rate (Rf) ---
print("\n--- CAPM Inputs ---")
dynamic_rf_rate = try_get_dynamic_rf(rf_proxy_ticker)
rf_prompt = f"Enter Risk-Free Rate (Rf) as percentage (e.g., '4.5' for 4.5%). Fetched {rf_proxy_ticker}?"
# ----> Use NEW default suggestion if dynamic fetch fails <----
suggested_rf_val = dynamic_rf_rate * 100 if dynamic_rf_rate is not None else DEFAULT_RF_RATE_SUGGESTION

risk_free_rate_input_pct = get_validated_float_input(
    rf_prompt, default_value=f"{suggested_rf_val:.2f}", allow_blank=False, min_val=0.0, max_val=20.0
)
risk_free_rate_input = risk_free_rate_input_pct / 100.0

input_sources['Rf'] = 'Dynamic Fetch (User Confirmed/Adjusted)' if dynamic_rf_rate is not None else f'User Input (Default: {DEFAULT_RF_RATE_SUGGESTION:.1f}%)' # Updated source string slightly
display(Markdown(f"**Using Risk-Free Rate (Rf):** {format_percent(risk_free_rate_input)} (`Source: {input_sources['Rf']}`)"))

# --- Get Expected Market Return (Rm) ---
suggested_mrp = DEFAULT_MARKET_RISK_PREMIUM
suggested_rm = risk_free_rate_input + suggested_mrp
rm_prompt = f"Enter Expected Market Return (Rm) as percentage (e.g., '10'). Suggestion based on Rf + {format_percent(suggested_mrp)} MRP is {format_percent(suggested_rm)}"

while True:
    market_return_input_pct = get_validated_float_input(
        rm_prompt, default_value=f"{suggested_rm * 100:.1f}", allow_blank=False, min_val=-10.0, max_val=35.0
    )
    market_return_input = market_return_input_pct / 100.0
    if market_return_input >= risk_free_rate_input: break
    else: print(f"❌ Error: Rm ({format_percent(market_return_input)}) must be >= Rf ({format_percent(risk_free_rate_input)}). Re-enter.")

market_risk_premium_input = market_return_input - risk_free_rate_input
input_sources['Rm'] = 'User Input'
input_sources['MRP'] = 'Derived'
display(Markdown(f"**Using Expected Market Return (Rm):** {format_percent(market_return_input)} (`Source: {input_sources['Rm']}`)"))
display(Markdown(f"**↳ Implied Market Risk Premium (MRP):** {format_percent(market_risk_premium_input)} (`Source: {input_sources['MRP']}`)"))

# --- Beta (Placeholder) ---
beta_input = None; input_sources['Beta'] = 'Pending'
# --- Credit Spread (Placeholder) ---
user_credit_spread = None; input_sources['Credit Spread'] = 'Pending'

# --- Get Industry Benchmark Ratios (with fallback defaults) ---
print("\n--- Industry Ratio Benchmarks ---")
display(Markdown("Enter benchmarks or leave blank for generic fallbacks. *(Find industry averages online)*"))
user_benchmarks = {}; benchmark_sources = {}
print("Provide benchmarks (e.g., ROE: enter '15' for 15%; P/E: enter '20'):")
for ratio_key in REQUIRED_BENCHMARK_RATIOS:
    default_val = DEFAULT_BENCHMARK_RATIOS.get(ratio_key, None)
    prompt = f"  Enter benchmark for {ratio_key}"
    is_percent_type = ratio_key in ['ROE', 'ROA', 'ROCE']
    min_b, max_b = (-50, 100) if is_percent_type else (-100, 500)

    bench_val_input = get_validated_float_input(
        prompt, default_value=f"{default_val:.2f}" if default_val is not None else None,
        allow_blank=True, min_val=min_b, max_val=max_b
    )

    if bench_val_input is not None:
        user_benchmarks[ratio_key] = bench_val_input
        benchmark_sources[ratio_key] = 'User Input'
        display_val = f"{bench_val_input:.1f}%" if is_percent_type else f"{bench_val_input:.2f}"
        display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: User`)"))
    elif default_val is not None:
         benchmark_sources[ratio_key] = 'Fallback Default'
         display_val = f"{default_val:.1f}%" if is_percent_type else f"{default_val:.2f}"
         display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: Default`)"))
    else:
        benchmark_sources[ratio_key] = 'Not Available'
        display(Markdown(f"  - **{ratio_key} Benchmark:** Not Provided / No Default"))
input_sources['Benchmarks'] = 'User Input or Default'

# %% [markdown]
# ## 4. Data Acquisition (Fetching Extended History)

# %%
stock_data = {}
try:
    stock = yf.Ticker(ticker_symbol)
    info = stock.info
    if not info or not info.get('regularMarketPrice'):
        print(f"Warning: Limited info for {ticker_symbol}. Trying history...")
        hist_check = stock.history(period="5d");
        if hist_check.empty: raise ValueError(f"Ticker '{ticker_symbol}' invalid/no data.")
        else: print(f"Warning: Proceeding with limited data."); info = {} if not isinstance(info, dict) else info
    stock_data['info'] = info

    # Standard statement fetching (usually 4 years)
    stock_data['financials'] = stock.financials
    stock_data['balance_sheet'] = stock.balance_sheet
    stock_data['cashflow'] = stock.cashflow
    if stock_data['financials'].empty or stock_data['balance_sheet'].empty or stock_data['cashflow'].empty:
       print(f"⚠️ Warning: Empty statements for {ticker_symbol}. DCF unreliable.")

    # Fetch recent history for charts/current price fallback
    stock_data['history'] = stock.history(period="2y")
    # ----> Fetch extended history for CAGR calculation (Still useful for historical analysis, even if not primary growth driver) <----
    stock_data['history_long'] = stock.history(period="11y") # Fetch 11 years for 10-year CAGR
    if stock_data['history_long'].empty:
        print(f"⚠️ Warning: Could not fetch long history (11y) for {ticker_symbol}.")


    company_name = info.get("longName", info.get("shortName", ticker_symbol))
    # Determine currency symbol, prioritizing Rupee for Indian stocks
    currency_code = info.get("currency", DEFAULT_CURRENCY_SYMBOL)
    if not isinstance(currency_code, str) or len(currency_code)>5 : currency_code=DEFAULT_CURRENCY_SYMBOL
    currency_symbol = "₹" if is_indian_stock else info.get("symbol", currency_code) # Prefer symbol if available, else code, force Rupee for Indian

    current_price = safe_get(info, ['currentPrice', 'regularMarketPrice', 'previousClose', 'lastPrice'], 'Info', default=np.nan)
    price_source = "yfinance Info"
    if not np.isfinite(current_price) and not stock_data['history'].empty:
        last_close_hist = stock_data['history']['Close'].iloc[-1]
        if pd.notna(last_close_hist): current_price = last_close_hist; price_source = "yfinance History"
    if not np.isfinite(current_price): print(f"⚠️ Warning: Could not determine current price."); current_price = np.nan

    shares_outstanding = safe_get(info, ['sharesOutstanding'], 'Info', default=0)

    beta_input = safe_get(info, ['beta'], 'Info', default=None)
    if beta_input is not None and np.isfinite(beta_input):
        input_sources['Beta'] = 'Dynamic Fetch'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))
    else:
         print(f"❌ ERROR: Beta not found dynamically or invalid.")
         beta_input = get_validated_float_input(f"Enter Beta for {ticker_symbol}:", allow_blank=False, min_val=-1.0, max_val=4.0)
         info['beta'] = beta_input; input_sources['Beta'] = 'User Input'
         display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))

    market_cap = safe_get(info, ['marketCap'], 'Info', default=0)
    mcap_source = "yfinance Info"
    if market_cap <= 0 and shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Calculated"; print(f"Info: Market Cap calculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}") # Pass flag
    elif market_cap <= 0: mcap_source = "Unavailable"; print(f"⚠️ Warning: Market Cap unavailable."); market_cap = np.nan

    sector = info.get("sector", "N/A"); industry = info.get("industry", "N/A")
    summary = info.get("longBusinessSummary", "No summary available.")

    display(Markdown(f"**Company:** {company_name}")); display(Markdown(f"**Sector:** {sector} | **Industry:** {industry}"))
    display(Markdown(f"**Currency:** {currency_code} ({currency_symbol})")); # Show both code and effective symbol
    # ----> Pass is_indian_stock flag to format_currency calls <----
    display(Markdown(f"**Current Market Price:** {format_currency(current_price, currency_symbol, is_indian=is_indian_stock)} (`Source: {price_source}`)"))
    display(Markdown(f"**Market Cap:** {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)} (`Source: {mcap_source}`)"))

except Exception as e: print(f"❌ CRITICAL ERROR fetching data: {e}"); sys.exit("Cannot proceed.")

if not isinstance(shares_outstanding, (int, float)) or shares_outstanding <= 0: sys.exit("Essential data missing: Shares Outstanding.")
if beta_input is None or not np.isfinite(beta_input): sys.exit("Essential data missing: Beta.")
if not np.isfinite(market_cap): print(f"⚠️ WARNING: Market Cap unavailable. WACC weights less accurate.")

# %% [markdown]
# ## 5. WACC Calculation

# %%
# WACC Function Definition (No changes needed here)
def calculate_wacc(stock_info_dict, financials_df, balance_sheet_df,
                   user_rf, user_mrp, user_beta, input_sources_dict,
                   currency_sym, is_indian): # Added currency params
    global user_credit_spread
    display(Markdown("### Calculating WACC..."))
    components = {}
    calculation_log = []
    calculation_log.append(f"Input Sources: Rf='{input_sources_dict.get('Rf')}', MRP='{input_sources_dict.get('MRP')}', Beta='{input_sources_dict.get('Beta')}'")

    beta_val = user_beta
    if beta_val <= 0: calculation_log.append(f"⚠️ Warning: Non-positive Beta ({beta_val:.2f}).")
    components['beta'] = beta_val; components['risk_free_rate'] = user_rf; components['market_risk_premium'] = user_mrp
    cost_of_equity = user_rf + beta_val * user_mrp
    components['cost_of_equity'] = cost_of_equity
    calculation_log.append(f"Inputs -> Rf: {format_percent(user_rf)}, Beta: {beta_val:.2f}, MRP: {format_percent(user_mrp)}")
    calculation_log.append(f"↳ Cost of Equity (Ke) [CAPM]: {format_percent(cost_of_equity)}")

    cost_of_debt_pre_tax = np.nan; tax_rate = np.nan; after_tax_cost_of_debt = np.nan
    total_debt = 0.0; debt_calculation_method = "N/A"; tax_rate_source = "N/A"

    if not balance_sheet_df.empty and not financials_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]; latest_fin = financials_df.iloc[:, 0]
        debt_keys = ['Total Debt', 'Long Term Debt', 'Long Term Debt And Capital Lease Obligation', 'Total Debt Non Current']
        long_term_debt = safe_get(latest_bs, debt_keys , 'Balance Sheet', default=0)
        current_debt_keys = ['Current Debt', 'Short Term Debt', 'Current Debt And Capital Lease Obligation', 'Short Long Term Debt', 'Total Debt Current']
        short_term_debt = safe_get(latest_bs, current_debt_keys , 'Balance Sheet', default=0)
        total_debt = long_term_debt + short_term_debt
        components['debt_value_used'] = total_debt
        interest_expense = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], 'Income Statement', default=0))

        implied_rate_calculated = False
        if total_debt > 1000 and interest_expense > 0:
             implied_rate = interest_expense / total_debt; implied_rate_calculated = True
             if MIN_COST_OF_DEBT <= implied_rate <= MAX_COST_OF_DEBT:
                  cost_of_debt_pre_tax = implied_rate; debt_calculation_method = "Implied";
                  # ----> Pass is_indian flag to format_currency calls <----
                  calculation_log.append(f"Interest Exp: {format_currency(interest_expense, symbol=currency_sym, is_indian=is_indian)}, Debt: {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
                  calculation_log.append(f"↳ Implied Cost of Debt (Kd pre-tax): {format_percent(cost_of_debt_pre_tax)}")
             else: calculation_log.append(f"⚠️ Implied Kd ({format_percent(implied_rate)}) out of bounds. Falling back.")
        elif total_debt > 1000: calculation_log.append(f"ℹ️ Found Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}) but no Interest Exp. Falling back.") # Pass flag

        if not debt_calculation_method.startswith("Implied"):
             if total_debt > 1000:
                 debt_calculation_method = "Default (Rf+Spread)"
                 if user_credit_spread is None:
                     spread_prompt = f"Implied Kd failed/NA. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
                     user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
                     user_credit_spread = user_credit_spread_pct / 100.0
                     input_sources_dict['Credit Spread'] = 'User Input (Default Kd)'
                 else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
                 cost_of_debt_pre_tax = user_rf + user_credit_spread
                 calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(cost_of_debt_pre_tax)}")
                 calculation_log.append(f"   (Spread Source: {input_sources_dict['Credit Spread']})")
             else:
                 debt_calculation_method = "Zero Debt"; cost_of_debt_pre_tax = 0.0; input_sources_dict['Credit Spread'] = 'N/A (Zero Debt)';
                 # ----> Pass is_indian flag to format_currency calls <----
                 calculation_log.append(f"ℹ️ Negligible Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}). Kd = 0%.")

        income_tax_expense = safe_get(latest_fin, ['Tax Provision', 'Income Tax Expense'], 'Income Statement', default=0)
        pretax_income = safe_get(latest_fin, ['Pretax Income', 'EBT', 'Earnings Before Tax'], 'Income Statement', default=0)
        if pretax_income <= 0 and income_tax_expense > 0: # Handle negative pretax income if tax was paid
             ebit_val = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=0); interest_exp_for_ebt = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], default=0))
             pretax_income_from_ebit = ebit_val - interest_exp_for_ebt
             if pretax_income_from_ebit > 0: pretax_income = pretax_income_from_ebit # Use calculated if positive
        if pretax_income > 0 and income_tax_expense >= 0:
            calculated_tax_rate = income_tax_expense / pretax_income; tax_rate = min(max(calculated_tax_rate, MIN_TAX_RATE), MAX_TAX_RATE); tax_rate_source = "Derived & Bounded"
            if abs(tax_rate - calculated_tax_rate) > 0.001: tax_rate_source = f"Derived ({format_percent(calculated_tax_rate)}), Bounded"
            # ----> Pass is_indian flag to format_currency calls <----
            calculation_log.append(f"Tax Exp: {format_currency(income_tax_expense, symbol=currency_sym, is_indian=is_indian)}, Pretax Inc: {format_currency(pretax_income, symbol=currency_sym, is_indian=is_indian)}")
            calculation_log.append(f"↳ Effective Tax Rate: {format_percent(tax_rate)} ({tax_rate_source})")
        else: tax_rate = DEFAULT_TAX_RATE; tax_rate_source = "Default"; calculation_log.append(f"⚠️ Cannot derive tax rate. Using default: {format_percent(tax_rate)}")
    else: # Financial statements empty
         calculation_log.append("⚠️ Financial statements empty. Using defaults."); debt_calculation_method = "Default (No Fin.)"; tax_rate_source = "Default (No Fin.)"
         if user_credit_spread is None:
              spread_prompt = f"No financials. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
              user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
              user_credit_spread = user_credit_spread_pct / 100.0; input_sources_dict['Credit Spread'] = 'User Input (No Fin.)'
         else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
         cost_of_debt_pre_tax = user_rf + user_credit_spread; tax_rate = DEFAULT_TAX_RATE; total_debt = 0; components['debt_value_used'] = 0
         calculation_log.append(f"Using Default Kd: {format_percent(cost_of_debt_pre_tax)} (Rf + Spread: {input_sources_dict['Credit Spread']})")
         calculation_log.append(f"Using Default Tax Rate: {format_percent(tax_rate)}")

    components['cost_of_debt'] = cost_of_debt_pre_tax if np.isfinite(cost_of_debt_pre_tax) else 0.0
    components['debt_calc_method'] = debt_calculation_method
    components['tax_rate'] = tax_rate if np.isfinite(tax_rate) else DEFAULT_TAX_RATE
    components['tax_rate_source'] = tax_rate_source
    if np.isfinite(cost_of_debt_pre_tax) and np.isfinite(tax_rate): after_tax_cost_of_debt = cost_of_debt_pre_tax * (1 - tax_rate)
    else: after_tax_cost_of_debt = components['cost_of_debt'] * (1 - DEFAULT_TAX_RATE) # Use default if calculation failed
    components['after_tax_cost_of_debt'] = after_tax_cost_of_debt
    if components['cost_of_debt'] > 0 : calculation_log.append(f"↳ Cost of Debt (After-tax): {format_percent(after_tax_cost_of_debt)}")

    market_cap_val = safe_get(stock_info_dict, ['marketCap'], 'Info', default=np.nan); equity_source = "Market Cap" if np.isfinite(market_cap_val) and market_cap_val > 0 else "N/A"
    equity_value_for_wacc = np.nan
    if np.isfinite(market_cap_val) and market_cap_val > 0: equity_value_for_wacc = market_cap_val
    else:
       calculation_log.append("⚠️ Market Cap unavailable/zero."); equity_source = "Book Equity Fallback"
       if not balance_sheet_df.empty:
           equity_keys = ['Total Stockholder Equity', 'Stockholders Equity', 'Total Equity Gross Minority Interest', 'Common Stock Equity']
           book_equity_val = safe_get(balance_sheet_df.iloc[:, 0], equity_keys, 'Balance Sheet', default=np.nan)
           if np.isfinite(book_equity_val) and book_equity_val > 1.0:
              equity_value_for_wacc = book_equity_val; calculation_log.append(f"Using Book Equity ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}) as proxy.") # Pass flag
           else: equity_source = "Placeholder (Book Invalid)"
       else: equity_source = "Placeholder (No BS)"
       if not np.isfinite(equity_value_for_wacc): equity_value_for_wacc = 1.0; calculation_log.append("⚠️ Market Cap & Book Equity unavailable. Using placeholder Equity Value (1.0). Weights inaccurate.")
    components['equity_value_used'] = equity_value_for_wacc; components['equity_source'] = equity_source

    equity_weight = np.nan; debt_weight = np.nan; weights_source = "N/A"; total_capital = equity_value_for_wacc + total_debt
    if np.isfinite(total_capital) and total_capital > 1.0:
        equity_weight = equity_value_for_wacc / total_capital; debt_weight = total_debt / total_capital; weights_source = "Calculated"
        total_weight = equity_weight + debt_weight # Normalize if calculation slightly off
        if abs(total_weight - 1.0) > 1e-6: equity_weight /= total_weight; debt_weight /= total_weight
    else: calculation_log.append(f"❌ ERROR: Total Capital ({format_currency(total_capital, symbol=currency_sym, is_indian=is_indian)}) invalid. Using default weights (70/30)."); equity_weight = 0.7; debt_weight = 0.3; weights_source = "Default" # Pass flag
    components['equity_weight'] = equity_weight; components['debt_weight'] = debt_weight; components['weights_source'] = weights_source
    # ----> Pass is_indian flag to format_currency calls <----
    calculation_log.append(f"Equity Value Used ({components['equity_source']}): {format_currency(components['equity_value_used'], symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"Debt Value Used ({components['debt_calc_method']}): {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"↳ Weights E/V: {format_percent(equity_weight)}, D/V: {format_percent(debt_weight)} (Source: {weights_source})")

    wacc = np.nan
    if all(np.isfinite(v) for v in [equity_weight, cost_of_equity, debt_weight, after_tax_cost_of_debt]):
        wacc = (equity_weight * cost_of_equity) + (debt_weight * after_tax_cost_of_debt); original_wacc = wacc; wacc = min(max(wacc, MIN_WACC), MAX_WACC)
        bound_note = ""
        if abs(wacc - original_wacc) > 0.0001 : bound_note = f"(Bounded from {format_percent(original_wacc)})"; calculation_log.append(f"⚠️ WACC ({format_percent(original_wacc)}) bounded.")
        calculation_log.append(f"▶️ Calculated WACC: {format_percent(wacc)} {bound_note}")
    else: calculation_log.append(f"❌ ERROR: Cannot calculate final WACC due to invalid components."); wacc = np.nan
    components['final_wacc'] = wacc

    display(Markdown("```\n" + "\n".join(calculation_log) + "\n```"))

    if np.isfinite(wacc) and wacc > 0 and np.isfinite(equity_weight) and np.isfinite(debt_weight):
        equity_contrib = equity_weight * cost_of_equity; debt_contrib = debt_weight * after_tax_cost_of_debt
        # Check if components reasonably sum up (handle potential float issues)
        if np.isfinite(equity_contrib) and np.isfinite(debt_contrib) and abs(equity_contrib + debt_contrib - wacc) < 0.01:
           labels = ['Cost of Equity Contribution', 'Cost of Debt Contribution']; values = [equity_contrib, debt_contrib]
           fig_wacc_donut = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4, marker_colors=px.colors.qualitative.Pastel, hovertemplate='<b>%{label}</b><br>Contribution: %{value:.2%}<extra></extra>', textinfo='percent', textfont_size=12, sort=False)])
           fig_wacc_donut.update_layout(title_text=f'WACC Composition (Total: {format_percent(wacc)})', title_x=0.5, showlegend=True, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), margin=dict(l=20, r=20, t=60, b=50), height=350)
           fig_wacc_donut.show()
        else: display(Markdown(f"*(WACC component contributions invalid/don't sum. Cannot generate chart.)*"))
    elif not np.isfinite(wacc): display(Markdown(f"*(WACC calculation failed. Cannot generate chart)*"))
    else: display(Markdown(f"*(WACC zero/negative or weights invalid. Cannot generate chart)*"))

    return wacc, components

# Execute WACC Calculation
# ----> Pass currency symbol and flag <----
discount_rate, wacc_components = calculate_wacc(
    stock_data['info'], stock_data['financials'], stock_data['balance_sheet'],
    risk_free_rate_input, market_risk_premium_input, beta_input, input_sources,
    currency_symbol, is_indian_stock
)

# %% [markdown]
# ## 6. FCFF Calculation & Projection (Initial Growth: 3yr Revenue CAGR)

# %%
# FCFF Function Definition (**NEW initial growth logic**)
def calculate_and_project_fcff(financials_df, balance_sheet_df, cashflow_df, history_long_df, # history_long kept for potential future use/analysis
                               wacc_comps, disc_rate, forecast_yrs,
                               final_user_rf,
                               default_init_growth, # This is 6.5% default from config
                               min_init_growth, # This is 1.1% floor from config
                               max_init_growth, # This is 15% ceiling from config
                               currency_sym, is_indian):
    display(Markdown("### Calculating & Projecting FCFF..."))
    projection_log = []
    projection_results = {
        'historical_fcffs': {}, # Store historical FCFFs used for growth calc
        'historical_revenues': {}, # Store historical revenues
        'current_fcff': np.nan, 'projected_fcffs': [], 'growth_rates': [],
        'pv_fcffs': [], 'cumulative_pv_fcff': np.nan, 'terminal_growth_rate': np.nan,
        'terminal_value': np.nan, 'pv_terminal_value': np.nan, 'error': None,
        'initial_growth_rate': np.nan, 'growth_source': 'N/A'
    }

    if financials_df.empty or balance_sheet_df.empty or cashflow_df.empty:
         projection_results['error'] = "Missing Financials/Balance Sheet/Cash Flow."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results
    if not np.isfinite(disc_rate):
         projection_results['error'] = "Invalid WACC provided."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results

    # --- Define Helper to calculate FCFF for a specific year column ---
    # (Same helper as before, ensures consistency)
    def get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, tax_rt):
        if not isinstance(fin_col, pd.Series) or not isinstance(bs_col, pd.Series) or not isinstance(cf_col, pd.Series):
            return np.nan, "Missing Statement Column"

        ebit = safe_get(fin_col, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
        depreciation_amortization = safe_get(cf_col, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
        capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
        capital_expenditures = abs(safe_get(cf_col, capex_keys, default=0))
        wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]
        change_in_wc = safe_get(cf_col, wc_keys, default=None)
        wc_src = "CF"

        if change_in_wc is None or not np.isfinite(change_in_wc):
            wc_src = "BS Calc"
            if isinstance(bs_prev_col, pd.Series): # Need previous year BS
                 curr_assets = safe_get(bs_col, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(bs_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                 prev_assets = safe_get(bs_prev_col, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(bs_prev_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                 if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val):
                     change_in_wc = curr_wc_val - prev_wc_val
                 else: change_in_wc = 0; wc_src = "BS Fail"
            else: change_in_wc = 0; wc_src = "No Prev BS"
        if not np.isfinite(change_in_wc): change_in_wc = 0

        fcff_val = np.nan; nopat_val = np.nan; calc_status = "OK"
        if np.isfinite(ebit) and np.isfinite(tax_rt):
            nopat_val = ebit * (1 - tax_rt)
            if np.isfinite(depreciation_amortization) and np.isfinite(capital_expenditures) and np.isfinite(change_in_wc):
                fcff_val = nopat_val + depreciation_amortization - capital_expenditures - change_in_wc
            else: calc_status = "Invalid Depr/CapEx/WC"
        else: calc_status = "Invalid EBIT/Tax"
        return fcff_val, calc_status
    # --- End FCFF Helper ---

    # --- Calculate Historical FCFF & Revenue (as far back as possible) ---
    max_hist_years = 4 # Need at least 4 years for 3-year CAGR
    num_statement_years = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    historical_fcffs_list = [] # Newest first (index 0 = T-0)
    historical_revenues_list = [] # Newest first

    effective_tax_rate = wacc_comps.get('tax_rate', DEFAULT_TAX_RATE) # Use consistent tax rate for historical calc

    projection_log.append("Calculating historical FCFF & Revenue...")
    # Limit loop to available data or required history for CAGR + 1 base year
    for i in range(min(num_statement_years, max_hist_years)):
        year_label = f"T-{i}"
        fin_col = financials_df.iloc[:, i]
        bs_col = balance_sheet_df.iloc[:, i]
        cf_col = cashflow_df.iloc[:, i]
        bs_prev_col = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None

        # Calculate FCFF for the year
        fcff_val, status = get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, effective_tax_rate)
        historical_fcffs_list.append(fcff_val if status == "OK" else np.nan) # Store NaN if calc failed

        # Get Revenue for the year
        rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
        revenue_val = safe_get(fin_col, rev_keys, default=np.nan)
        historical_revenues_list.append(revenue_val)

    projection_results['historical_fcffs'] = {f"T-{i}": v for i, v in enumerate(historical_fcffs_list)}
    projection_results['historical_revenues'] = {f"T-{i}": v for i, v in enumerate(historical_revenues_list)}
    projection_results['current_fcff'] = historical_fcffs_list[0] if historical_fcffs_list and np.isfinite(historical_fcffs_list[0]) else np.nan
    fcff_current = projection_results['current_fcff']

    # --- Display Calculation Log for Current FCFF (if calculated) ---
    if np.isfinite(fcff_current):
        latest_fin = financials_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]
        ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
        nopat = ebit * (1-effective_tax_rate) if np.isfinite(ebit) else np.nan
        depreciation_amortization = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
        capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
        capital_expenditures = abs(safe_get(latest_cf, capex_keys, default=0))
        wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]
        change_in_wc = safe_get(latest_cf, wc_keys, default=None)
        wc_source = "Direct from Cash Flow"
        if change_in_wc is None or not np.isfinite(change_in_wc):
            if balance_sheet_df.shape[1] >= 2:
                prev_bs = balance_sheet_df.iloc[:, 1]
                curr_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                prev_assets = safe_get(prev_bs, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(prev_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val): change_in_wc = curr_wc_val - prev_wc_val; wc_source = "Calculated from BS"
                else: change_in_wc = 0; wc_source = "Calc Failed (Invalid BS)"
            else: change_in_wc = 0; wc_source = "Default (No Prev BS)"
        if not np.isfinite(change_in_wc): change_in_wc = 0
        projection_log.append(f"EBIT (T-0): {format_currency(ebit, currency_sym, is_indian=is_indian)}, Tax Rate ({wacc_comps.get('tax_rate_source','?')}) {format_percent(effective_tax_rate)} -> NOPAT: {format_currency(nopat, currency_sym, is_indian=is_indian)}")
        projection_log.append(f"  + Dep&Amort: {format_currency(depreciation_amortization, currency_sym, is_indian=is_indian)}")
        projection_log.append(f"  - CapEx: {format_currency(capital_expenditures, currency_sym, is_indian=is_indian)}")
        projection_log.append(f"  - Change WC ({wc_source}): {format_currency(change_in_wc, currency_sym, is_indian=is_indian)}")
        projection_log.append(f"▶️ Historical FCFF (T-0): {format_currency(fcff_current, currency_sym, is_indian=is_indian)}")
    else:
        projection_log.append("⚠️ Could not calculate current historical FCFF (T-0).")

    if not np.isfinite(fcff_current):
        projection_results['error'] = "Invalid historical FCFF (T-0). Cannot proceed.";
        display(Markdown(f"❌ ERROR: {projection_results['error']}")); display(Markdown("```\n" + "\n".join(projection_log) + "\n```")); return projection_results

    # --- NEW: Initial Growth Rate Calculation (3yr Revenue CAGR) ---
    projection_log.append(f"--- Determining Initial Growth Rate (3yr Revenue CAGR Focus | Bounds: {format_percent(min_init_growth)}-{format_percent(max_init_growth)}) ---")
    initial_growth_rate = np.nan
    growth_source = "N/A"
    raw_growth_rate = np.nan
    source_detail = ""

    # 1. Attempt 3-Year Revenue CAGR
    if len(historical_revenues_list) >= 4: # Need T-0, T-1, T-2, T-3 data points
        end_val = historical_revenues_list[0]   # Revenue T-0 (Most Recent)
        start_val = historical_revenues_list[3] # Revenue T-3
        projection_log.append(f"  Attempting 3yr Revenue CAGR (Start T-3: {format_currency(start_val, currency_sym, 0, is_indian)}, End T-0: {format_currency(end_val, currency_sym, 0, is_indian)})")
        calculated_cagr = calculate_cagr(start_val, end_val, 3)

        if pd.notna(calculated_cagr) and math.isfinite(calculated_cagr):
            raw_growth_rate = calculated_cagr
            source_detail = "3yr Revenue CAGR"
            projection_log.append(f"  ✅ Success: Calculated {source_detail} = {format_percent(raw_growth_rate)}")
        else:
            projection_log.append(f"  - 3yr Revenue CAGR: Calculation failed or invalid data (Start={start_val}, End={end_val}). Falling back to default.")
            source_detail = f"Default ({format_percent(default_init_growth)} - Rev CAGR Failed)"
            raw_growth_rate = default_init_growth # Use the CONFIG default (6.5%)
    else:
        projection_log.append(f"  - 3yr Revenue CAGR: Insufficient historical Revenue data ({len(historical_revenues_list)} points available, need 4). Falling back to default.")
        source_detail = f"Default ({format_percent(default_init_growth)} - No 3yr Rev Data)"
        raw_growth_rate = default_init_growth # Use the CONFIG default (6.5%)

    # 2. Apply Ceilings and Floors (New Rules: 15% upper, 1.1% lower)
    bounded_growth = raw_growth_rate
    bounding_applied = ""

    if bounded_growth > max_init_growth:
        bounded_growth = max_init_growth
        bounding_applied = " (Ceiling Applied)"
        projection_log.append(f"  ℹ️ Applying Upper Ceiling: {format_percent(max_init_growth)}")

    if bounded_growth < min_init_growth:
        bounded_growth = min_init_growth
        # Append or overwrite bounding info? Append is clearer.
        if bounding_applied:
            bounding_applied += " & Floor Applied"
        else:
            bounding_applied = " (Floor Applied)"
        projection_log.append(f"  ℹ️ Applying Lower Ceiling: {format_percent(min_init_growth)}")

    initial_growth_rate = bounded_growth
    growth_source = source_detail + bounding_applied

    projection_log.append(f"▶️ Using Final Initial Growth Rate: {format_percent(initial_growth_rate)} (Source: {growth_source})")
    projection_results['initial_growth_rate'] = initial_growth_rate
    projection_results['growth_source'] = growth_source
    # --- END NEW Initial Growth Logic ---


    # Terminal Growth Rate Calculation (Remains the same conservative approach)
    # Use MIN_GROWTH_FLOOR (0.0%) here for terminal rate floor
    g_cap_initial_valid = initial_growth_rate if pd.notna(initial_growth_rate) and math.isfinite(initial_growth_rate) else TERMINAL_GROWTH_RATE_CAP # Fallback if initial is invalid

    g_cap_config = TERMINAL_GROWTH_RATE_CAP; g_cap_wacc = disc_rate * 0.67; g_cap_rf = final_user_rf * 0.95 # Use final RF
    projection_log.append(f"Terminal Growth Caps Considered: Config={format_percent(g_cap_config)}, WACC*0.67={format_percent(g_cap_wacc)}, Initial={format_percent(g_cap_initial_valid)}, Rf*0.95={format_percent(g_cap_rf)}")
    terminal_growth_rate = min(g_cap_config, g_cap_wacc, g_cap_initial_valid, g_cap_rf); projection_log.append(f"↳ Minimum of Caps: {format_percent(terminal_growth_rate)}")
    original_g_before_floor = terminal_growth_rate; terminal_growth_rate = max(MIN_GROWTH_FLOOR, terminal_growth_rate) # Apply floor (0.0%)
    if abs(terminal_growth_rate - original_g_before_floor) > 1e-6 : projection_log.append(f"↳ Adjusted by Floor ({format_percent(MIN_GROWTH_FLOOR)}): {format_percent(terminal_growth_rate)}")
    if terminal_growth_rate >= disc_rate: original_g = terminal_growth_rate; terminal_growth_rate = disc_rate - 0.005; projection_log.append(f"⚠️ Terminal Growth ({format_percent(original_g)}) >= WACC ({format_percent(disc_rate)}). Reduced to {format_percent(terminal_growth_rate)}.")
    projection_log.append(f"▶️ Using Final Terminal Growth Rate (g): {format_percent(terminal_growth_rate)}"); projection_results['terminal_growth_rate'] = terminal_growth_rate

    # Project FCFFs using Linear Decay from initial_growth_rate -> terminal_growth_rate
    projected_fcffs = []; growth_rates_applied = []; pv_fcffs = []; cumulative_pv_fcff = 0.0
    current_fcff_for_proj = fcff_current # Start projection base from T-0 FCFF

    # Ensure effective initial growth is valid, fallback if needed (shouldn't happen with new logic, but safeguard)
    if pd.isna(initial_growth_rate) or not math.isfinite(initial_growth_rate):
        projection_log.append(f"⚠️ Invalid initial growth rate ({initial_growth_rate}) determined. Using default {format_percent(default_init_growth)} for projection start.")
        effective_initial_growth = default_init_growth
    else:
        effective_initial_growth = initial_growth_rate

    effective_terminal_growth = terminal_growth_rate

    # Handle negative starting FCFF carefully
    if current_fcff_for_proj <= 0:
        projection_log.append("⚠️ Starting FCFF (T-0) <= 0. Applying minimal growth.")
        # Still allow the calculated initial growth, but cap it minimally positive if it was higher
        effective_initial_growth = min(effective_initial_growth, 0.01);
        effective_terminal_growth = min(terminal_growth_rate, 0.00) # Minimal non-negative terminal

    projection_log.append("--- FCFF Projections ---"); projection_log.append(f"{'Year':<5} {'Growth':<8} {'FCFF':<18} {'PV Factor':<10} {'PV FCFF':<18}"); projection_log.append("-" * 60) # Adjusted width
    last_valid_fcff = current_fcff_for_proj # Use the starting FCFF (T-0) for year 1 growth calculation

    for year in range(1, forecast_yrs + 1):
        # Linear decay of growth from initial to terminal
        decay_factor = (forecast_yrs - year) / forecast_yrs # Starts closer to initial, ends at terminal
        growth_rate = effective_terminal_growth + (effective_initial_growth - effective_terminal_growth) * decay_factor
        # Apply floor (0%) to yearly growth during projection phase
        growth_rate = max(MIN_GROWTH_FLOOR, growth_rate)

        # If started negative, constrain yearly growth rate more tightly to avoid exploding negativity/positivity unrealistically
        if current_fcff_for_proj <= 0:
            growth_rate = max(-0.10, min(growth_rate, 0.05)) # Limit yearly growth rate range if starting FCFF was negative

        growth_rates_applied.append(growth_rate)
        fcff_forecast = last_valid_fcff * (1 + growth_rate) # Calculate next FCFF

        if not np.isfinite(fcff_forecast): fcff_forecast = 0 # Handle potential NaN
        projected_fcffs.append(fcff_forecast)
        last_valid_fcff = fcff_forecast # IMPORTANT: Update last_valid_fcff for the *next* iteration

        pv_factor = 1 / ((1 + disc_rate) ** year); pv_fcff = fcff_forecast * pv_factor
        if not np.isfinite(pv_fcff): pv_fcff = 0 # Handle potential NaN PV
        pv_fcffs.append(pv_fcff); cumulative_pv_fcff += pv_fcff # Add PV (even if 0)
        projection_log.append(f"{year:<5} {format_percent(growth_rate):<8} {format_currency(fcff_forecast, '', 0, is_indian=is_indian):<18} {pv_factor:<10.4f} {format_currency(pv_fcff, '', 0, is_indian=is_indian):<18}")

    projection_results['projected_fcffs'] = projected_fcffs; projection_results['growth_rates'] = growth_rates_applied; projection_results['pv_fcffs'] = pv_fcffs
    projection_results['cumulative_pv_fcff'] = cumulative_pv_fcff if np.isfinite(cumulative_pv_fcff) else 0.0 # Handle NaN sum
    projection_log.append("-" * 60); projection_log.append(f"Sum of PV(Forecasted FCFFs): {format_currency(cumulative_pv_fcff, symbol=currency_sym, is_indian=is_indian)}") # Use proper symbol here

    # Calculate Terminal Value (Remains the same)
    terminal_value = np.nan; pv_terminal_value = np.nan
    final_forecast_fcff = projected_fcffs[-1] if projected_fcffs and np.isfinite(projected_fcffs[-1]) else 0

    if (disc_rate - terminal_growth_rate) <= 1e-6 or not np.isfinite(terminal_growth_rate):
         projection_log.append(f"❌ ERROR: WACC ({format_percent(disc_rate)}) <= Terminal Growth ({format_percent(terminal_growth_rate)}) or g invalid. TV calc failed."); projection_results['error'] = "WACC <= g or g invalid. TV Calc failed."
    elif not np.isfinite(final_forecast_fcff):
         projection_log.append(f"❌ ERROR: Final projected FCFF (Year {forecast_yrs}) is invalid. TV calc failed."); projection_results['error'] = "Final FCFF invalid. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    else:
         if final_forecast_fcff <= 0:
             projection_log.append(f"ℹ️ FCFF Year {forecast_yrs} ({format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)}) <= 0. Setting TV=0."); # Pass flag
             terminal_value = 0.0; pv_terminal_value = 0.0
         else:
              fcff_n_plus_1 = final_forecast_fcff * (1 + terminal_growth_rate)
              terminal_value = fcff_n_plus_1 / (disc_rate - terminal_growth_rate)
              if not np.isfinite(terminal_value) or terminal_value < 0:
                  projection_log.append(f"⚠️ Calculated TV invalid/negative ({format_currency(terminal_value, currency_sym, is_indian=is_indian)}). Setting TV=0.") # Check validity post-calc
                  terminal_value = 0.0; pv_terminal_value = 0.0
              else:
                   pv_factor_tv = 1 / ((1 + disc_rate) ** forecast_yrs); pv_terminal_value = terminal_value * pv_factor_tv
                   projection_log.append(f"FCFF Yr {forecast_yrs}: {format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)} -> FCFF Yr {forecast_yrs+1}: {format_currency(fcff_n_plus_1, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"Terminal Value (Yr {forecast_yrs}): {format_currency(terminal_value, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"PV Factor TV: {pv_factor_tv:.4f} -> PV(TV): {format_currency(pv_terminal_value, currency_sym, is_indian=is_indian)}")

    projection_results['terminal_value'] = terminal_value if np.isfinite(terminal_value) else 0.0
    projection_results['pv_terminal_value'] = pv_terminal_value if np.isfinite(pv_terminal_value) else 0.0 # Handle final NaN

    display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
    return projection_results

# Execute FCFF calculation and projection
if np.isfinite(discount_rate):
    # ----> Pass currency symbol, flag, NEW growth parameters <----
    fcff_results = calculate_and_project_fcff(
        stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'], stock_data['history_long'], # Pass history_long (still potentially useful)
        wacc_components, discount_rate, FORECAST_YEARS, risk_free_rate_input,
        DEFAULT_INITIAL_GROWTH_RATE, # Pass the 6.5% default
        MIN_INITIAL_GROWTH_RATE,     # Pass the 1.1% floor
        MAX_INITIAL_GROWTH_RATE,     # Pass the 15% ceiling
        currency_symbol, is_indian_stock)
else:
    display(Markdown("❌ **Skipping FCFF Projection due to invalid WACC.**")); fcff_results = {'error': 'Invalid WACC.'}


# %% [markdown]
# ## 7. Enterprise Value and Equity Value Calculation

# %%
# Valuation Function Definition (No changes needed here)
def calculate_final_valuation(fcff_data, balance_sheet_df, shares_outstanding_val, currency_sym, wacc_data, is_indian): # Added flag
    display(Markdown("### Final Valuation Calculation..."))
    valuation = {'enterprise_value': np.nan, 'equity_value': np.nan, 'intrinsic_value_per_share': np.nan, 'error': None}
    valuation_log = []
    if fcff_data is None or fcff_data.get('error'):
        valuation['error'] = f"FCFF Error: {fcff_data.get('error', 'Unknown')}"; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    pv_forecast_sum = fcff_data.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_data.get('pv_terminal_value', np.nan)
    if not np.isfinite(pv_forecast_sum) or not np.isfinite(pv_tv):
         valuation['error'] = f"Invalid PV components (PV FCFF: {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}, PV TV: {format_currency(pv_tv, currency_sym, is_indian=is_indian)})."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation # Pass flag

    enterprise_value = pv_forecast_sum + pv_tv; valuation['enterprise_value'] = enterprise_value
    # ----> Pass is_indian flag to format_currency calls <----
    valuation_log.append(f"PV(Forecast FCFFs): {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"PV(Terminal Value):   {format_currency(pv_tv, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"↳ Enterprise Value (EV): {format_currency(enterprise_value, currency_sym, is_indian=is_indian)}")

    # Get Debt & Cash from BS (use Debt from WACC for consistency)
    total_debt_latest = 0.0; cash_and_equivalents = 0.0; minority_interest = 0.0
    debt_source_info = wacc_data.get('debt_calc_method', '?'); total_debt_latest = wacc_data.get('debt_value_used', 0.0) # Get debt used in WACC

    if not balance_sheet_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        cash_keys = ["Cash And Cash Equivalents", "Cash Cash Equivalents And Short Term Investments", "Cash Financial", "Cash", "Cash And Short Term Investments"]
        cash_and_equivalents = safe_get(latest_bs, cash_keys, "BS", default=0)
        valuation_log.append(f"  + Cash & Equiv (BS): {format_currency(cash_and_equivalents, currency_sym, is_indian=is_indian)}")
        minority_interest = safe_get(latest_bs, ["Minority Interest", "Noncontrolling Interest"], "BS", default=0)
        if minority_interest != 0.0:
            valuation_log.append(f"  - Minority Interest: {format_currency(minority_interest, currency_sym, is_indian=is_indian)}")
    else:
        valuation_log.append("⚠️ Balance Sheet empty.")
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}") # Still use WACC debt if available
        valuation_log.append(f"  + Cash & Equiv: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")
        valuation_log.append(f"  - Minority Interest: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")

    if not all(np.isfinite(v) for v in [enterprise_value, total_debt_latest, cash_and_equivalents, minority_interest]):
         valuation['error'] = "Invalid component for EV->Equity bridge."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    equity_value = enterprise_value - total_debt_latest + cash_and_equivalents - minority_interest; valuation['equity_value'] = equity_value
    # ----> Pass is_indian flag to format_currency calls <----
    valuation_log.append(f"↳ Equity Value: {format_currency(equity_value, currency_sym, is_indian=is_indian)}")

    # Calculate Intrinsic Value Per Share
    intrinsic_value_per_share = np.nan
    if not np.isfinite(equity_value): valuation['error'] = f"Equity value invalid ({equity_value})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    elif shares_outstanding_val <= 0: valuation['error'] = f"Shares outstanding invalid ({shares_outstanding_val})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    else:
        intrinsic_value_per_share = equity_value / shares_outstanding_val;
        # Format shares with commas, independent of currency
        valuation_log.append(f"  ÷ Shares Out: {shares_outstanding_val:,.0f}")
        # ----> Pass is_indian flag to format_currency calls <----
        valuation_log.append(f"▶️ Intrinsic Value/Share: {format_currency(intrinsic_value_per_share, currency_sym, is_indian=is_indian)}")
    valuation['intrinsic_value_per_share'] = intrinsic_value_per_share if np.isfinite(intrinsic_value_per_share) else np.nan

    display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
    return valuation

# Execute Final Valuation
if 'error' not in fcff_results or fcff_results['error'] is None:
    # ----> Pass flag <----
    final_valuation = calculate_final_valuation(fcff_results, stock_data['balance_sheet'], shares_outstanding, currency_symbol, wacc_components, is_indian_stock)
else:
    display(Markdown("❌ **Skipping Final Valuation due to FCFF errors.**")); final_valuation = {'error': fcff_results.get('error', 'Upstream FCFF error.')}


# %% [markdown]
# ## 8. Financial Ratio Analysis

# %%
# Ratio Function Definition (No changes needed here)
def analyze_ratios(stock_info_dict, financials_df, balance_sheet_df, cashflow_df,
                   enterprise_val, current_price_val, shares_outstanding_val,
                   currency_sym, user_benchmark_dict, # Only user inputs
                   benchmark_source_dict, # Dict tracking source per ratio
                   wacc_data, is_indian): # Pass WACC data for consistent debt and flag

    """ Calculates ratios, compares to user/fallback benchmarks, uses WACC-consistent debt. """
    display(Markdown("### Financial Ratio Analysis..."))
    ratios = {} # Calculated ratios {RatioName: value}
    ratio_log = []
    ratio_data_for_table = [] # For HTML table display
    ratio_plotly_data = {'Ratio': [], 'Company': [], 'Benchmark': [], 'Benchmark Source': [], 'Status': []} # For Plotly chart
    company_nm = stock_info_dict.get("shortName", ticker_symbol)

    # --- Pre-checks ---
    valid_price = current_price_val is not None and np.isfinite(current_price_val) and current_price_val > 0
    valid_shares = shares_outstanding_val is not None and np.isfinite(shares_outstanding_val) and shares_outstanding_val > 0
    valid_enterprise_val = enterprise_val is not None and np.isfinite(enterprise_val)

    # Check for essential statements
    if financials_df.empty or balance_sheet_df.empty:
         ratio_log.append("❌ ERROR: Income Statement or Balance Sheet empty. Cannot calculate most ratios.")
         display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
         return {}, [] # Return empty results

    latest_fin = financials_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0] if not cashflow_df.empty else pd.Series(dtype=float)

    # --- Calculate Ratio Components (using safe_get with NaN default where appropriate) ---
    net_income = safe_get(latest_fin, ["Net Income Continuous Operations", "Net Income", "Net Income Common Stockholders", "Net Income Applicable To Common Shares"], default=np.nan)
    revenue = safe_get(latest_fin, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
    depreciation = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0) # Default 0 ok here
    ebitda = (ebit + depreciation) if np.isfinite(ebit) and np.isfinite(depreciation) else np.nan

    total_equity = safe_get(latest_bs, ["Total Stockholder Equity", "Stockholders Equity", "Total Equity Gross Minority Interest", "Common Stock Equity"], default=np.nan)
    total_assets = safe_get(latest_bs, ["Total Assets"], default=np.nan)
    current_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=np.nan)
    current_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=np.nan)
    inventory = safe_get(latest_bs, ["Inventory"], default=0) # Default 0 ok here

    # --- Use Debt value consistent with WACC calculation ---
    total_debt_ratio = wacc_data.get('debt_value_used', 0.0); debt_source_info = wacc_data.get('debt_calc_method', '?');
    # ----> Pass is_indian flag to format_currency calls <----
    ratio_log.append(f"Using Debt = {format_currency(total_debt_ratio, currency_sym, is_indian=is_indian)} (Source: WACC - {debt_source_info})")

    # Per Share Metrics
    eps = (net_income / shares_outstanding_val) if valid_shares and np.isfinite(net_income) and net_income != 0 else np.nan
    bvps = (total_equity / shares_outstanding_val) if valid_shares and np.isfinite(total_equity) and total_equity != 0 else np.nan

    # Capital Employed (ROCE denominator)
    capital_employed = np.nan; ce_source = "N/A"
    # Definition 1: Total Assets - Current Liabilities
    if np.isfinite(total_assets) and np.isfinite(current_liab):
        capital_employed_def1 = total_assets - current_liab
        # Check if calculated value is valid (positive)
        if np.isfinite(capital_employed_def1) and capital_employed_def1 > 1.0: # Use 1.0 as a small threshold
            capital_employed = capital_employed_def1
            ce_source = "A-CL"

    # Definition 2: Total Equity + Total Debt (Fallback)
    if not np.isfinite(capital_employed): # If Def1 failed or was invalid, try Def2
        if np.isfinite(total_equity) and np.isfinite(total_debt_ratio):
             capital_employed_def2 = total_equity + total_debt_ratio
             if np.isfinite(capital_employed_def2) and capital_employed_def2 > 1.0: # Use 1.0 threshold
                 capital_employed = capital_employed_def2
                 ce_source = "E+D" # Update source if fallback used

    # Check if still invalid after fallbacks
    if not np.isfinite(capital_employed):
         ce_source = "Invalid/Unavailable"
         ratio_log.append("⚠️ Capital Employed could not be reliably calculated. ROCE affected.")

    # --- Calculate Ratios (Result is NaN if components invalid/zero denominator) ---
    ratios['P/E'] = (current_price_val / eps) if valid_price and np.isfinite(eps) and eps != 0 else np.nan
    ratios['P/B'] = (current_price_val / bvps) if valid_price and np.isfinite(bvps) and bvps != 0 else np.nan
    ratios['EV/EBITDA'] = (enterprise_val / ebitda) if valid_enterprise_val and np.isfinite(ebitda) and ebitda != 0 else np.nan
    ratios['ROE'] = (net_income / total_equity) if np.isfinite(net_income) and np.isfinite(total_equity) and total_equity != 0 else np.nan
    ratios['ROA'] = (net_income / total_assets) if np.isfinite(net_income) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['ROCE'] = (ebit / capital_employed) if np.isfinite(ebit) and np.isfinite(capital_employed) and capital_employed != 0 else np.nan # Uses calculated CE
    ratios['Asset Turnover'] = (revenue / total_assets) if np.isfinite(revenue) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['Current Ratio'] = (current_assets / current_liab) if np.isfinite(current_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan
    quick_assets = (current_assets - inventory) if np.isfinite(current_assets) and np.isfinite(inventory) else np.nan
    ratios['Quick Ratio'] = (quick_assets / current_liab) if np.isfinite(quick_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan

    # --- Compare with User/Fallback Benchmarks & Prepare Data Tables ---
    display(Markdown("*(Comparing ratios to user/fallback benchmarks...)*"))
    for key in REQUIRED_BENCHMARK_RATIOS:
        company_value = ratios.get(key); is_percent_type = key in ['ROE', 'ROA', 'ROCE']
        # Handle formatting, including potential Inf values
        # NOTE: Ratios themselves are typically NOT formatted as currency (Lakh/Crore)
        # So we use format_currency with '' symbol and is_indian=False intentionally here
        value_fmt = format_percent(company_value, 1) if is_percent_type else format_currency(company_value, '', 2, is_indian=False)
        if value_fmt == "N/A" and company_value is not None and not math.isfinite(company_value): value_fmt = "Inf" # Explicitly show Inf

        # Determine which benchmark to use and its source
        user_benchmark = user_benchmark_dict.get(key, None); fallback_benchmark = DEFAULT_BENCHMARK_RATIOS.get(key, None)
        benchmark_source = benchmark_source_dict.get(key, "?"); final_benchmark = None
        if benchmark_source == 'User Input' and user_benchmark is not None: final_benchmark = user_benchmark
        elif benchmark_source == 'Fallback Default' and fallback_benchmark is not None: final_benchmark = fallback_benchmark
        else: benchmark_source = "Not Available" # If neither user nor default was set

        # Format benchmark for display
        bench_fmt = "N/A"
        if final_benchmark is not None: bench_fmt = f"{final_benchmark:.1f}%" if is_percent_type else f"{final_benchmark:.2f}"; bench_fmt += f" <i style='font-size:0.8em; color:grey'>({benchmark_source.replace('Fallback ','')})</i>"
        else: bench_fmt = "<i style='color:grey'>Not Provided</i>"

        # Determine status vs benchmark
        status = "<i style='color:grey'>N/A</i>"; status_color = "grey"
        if np.isfinite(company_value) and final_benchmark is not None and np.isfinite(final_benchmark):
            numeric_value = company_value; numeric_benchmark = final_benchmark
            good_color = '#2ecc71'; bad_color = '#e74c3c'; neutral_color = '#f39c12'; grey_color='#7f8c8d'
            # Handle potential division by zero for ratio diff
            if abs(numeric_benchmark) > 1e-9: ratio_diff = (numeric_value - numeric_benchmark) / abs(numeric_benchmark)
            else: ratio_diff = float('inf') if numeric_value > 0 else float('-inf') if numeric_value < 0 else 0.0
            # Set status based on ratio type and difference threshold (e.g., +/- 20%)
            if key in ['ROE', 'ROA', 'ROCE', 'Asset Turnover', 'Current Ratio', 'Quick Ratio']: # Higher is better
                 if ratio_diff > 0.20: status_color = good_color; status = f"⬆️ Strong"
                 elif ratio_diff < -0.20: status_color = bad_color; status = f"⬇️ Weak"
                 else: status_color = grey_color; status = f"~ Average"
            elif key in ['P/E', 'P/B', 'EV/EBITDA']: # Lower might be better (valuation) or neutral
                 if ratio_diff < -0.20: status_color = good_color; status = f"✅ Low"
                 elif ratio_diff > 0.20: status_color = neutral_color; status = f"⚠️ High" # High valuation isn't always bad, but notable
                 else: status_color = grey_color; status = f"~ Average"
            status = f"<span style='color:{status_color}'>{status}</span>"
            # Prepare data for Plotly chart (convert % ratios back to % for plotting if needed)
            plot_value = numeric_value * 100 if is_percent_type else numeric_value; plot_benchmark = numeric_benchmark # Benchmark already in correct format
            ratio_plotly_data['Ratio'].append(key); ratio_plotly_data['Company'].append(plot_value); ratio_plotly_data['Benchmark'].append(plot_benchmark);
            ratio_plotly_data['Benchmark Source'].append(benchmark_source); ratio_plotly_data['Status'].append(status) # Store HTML status for potential hover
        elif not np.isfinite(company_value): status = "<i style='color:grey'>Company N/A</i>"
        elif final_benchmark is None: status = "<i style='color:grey'>No Benchmark</i>"
        ratio_data_for_table.append([f"<b>{key}</b>", value_fmt, bench_fmt, status])

    # --- Display Logs & Table ---
    if ratio_log: display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
    if ratio_data_for_table:
        header = ["Ratio", f"{company_nm} Value", "Benchmark (Source)", "Status"]; html_table = "<p>(*Status vs User/Default benchmarks. Threshold: +/- 20%.)</p>"; html_table += "<table border='1' style='border-collapse:collapse; width:98%; margin-left:auto; margin-right:auto; font-size: 0.9em;'>"; html_table += "<thead style='background-color:#e9ecef; color: #495057;'><tr>" + "".join(f"<th style='padding: 8px; text-align:center;'>{h}</th>" for h in header) + "</tr></thead>"; html_table += "<tbody>"
        for i, row in enumerate(ratio_data_for_table): bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff"; html_table += f"<tr style='background-color:{bg_color};'>"; html_table += f"<td style='padding: 6px; text-align:left;'>{row[0]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[1]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[2]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[3]}</td>"; html_table += "</tr>"
        html_table += "</tbody></table><br>"; display(HTML(html_table))
    else: display(Markdown("*No ratio data available.*"))

    # --- Display Ratio Comparison Chart (Plotly Bullet) ---
    if ratio_plotly_data['Ratio']:
       num_ratios = len(ratio_plotly_data['Ratio']); fig_ratios = go.Figure()
       def get_range_colors(k): return ["#fadbd8", "#fdebd0", "#d5f5e3"] if k in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio'] else ["#d5f5e3","#fdebd0","#fadbd8"] if k in ['P/E','P/B','EV/EBITDA'] else ['lightgrey','darkgrey','grey'] # Bad/Ok/Good vs Good/Ok/Bad
       for i in range(num_ratios):
           ratio_name = ratio_plotly_data['Ratio'][i]; company_val = ratio_plotly_data['Company'][i]; benchmark_val = ratio_plotly_data['Benchmark'][i]; bench_src_plot = ratio_plotly_data['Benchmark Source'][i]
           # Skip plotting if data invalid
           if not np.isfinite(company_val) or not np.isfinite(benchmark_val): print(f"Skipping plot for {ratio_name}: Invalid Value ({company_val}) or Bench ({benchmark_val})."); continue

           range_colors = get_range_colors(ratio_name); is_percent = ratio_name in ['ROE', 'ROA', 'ROCE']
           # Define poor/good factors relative to benchmark (handle zero/negative benchmarks)
           poor_f=0.8; good_f=1.2;
           range1_limit = benchmark_val * (poor_f if benchmark_val >=0 else good_f) # Lower band for +ve, upper for -ve
           range2_limit = benchmark_val * (good_f if benchmark_val >=0 else poor_f) # Upper band for +ve, lower for -ve
           # Dynamic axis range setting
           abs_max = max(abs(company_val), abs(benchmark_val)); plot_max_val = abs_max * 1.7 + 1; plot_min_val = 0
           if company_val < 0 or benchmark_val < 0: plot_min_val = min(company_val, benchmark_val) * 1.7 - 1 # Extend lower if negative values present
           # Apply some sensible caps to axis ranges based on ratio type
           if ratio_name in ['P/E', 'EV/EBITDA']: plot_max_val = min(plot_max_val, 100); plot_min_val = max(plot_min_val, -20)
           elif ratio_name in ['P/B']: plot_max_val = min(plot_max_val, 20); plot_min_val = max(plot_min_val, -5)
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: plot_max_val = min(plot_max_val, 5); plot_min_val = max(plot_min_val, 0)
           elif is_percent: plot_max_val = min(plot_max_val, 75); plot_min_val= max(plot_min_val,-50)
           else: plot_max_val = min(plot_max_val, 1000) # Generic cap
           if plot_min_val >= plot_max_val: plot_max_val = plot_min_val + 1 # Ensure max > min

           # Define gauge steps based on calculated ranges
           steps = [];
           if abs(benchmark_val) < 1e-9: steps = [{'range': [plot_min_val, plot_max_val], 'color':'lightgrey'}] # Handle zero benchmark case
           else: step_ranges = sorted([plot_min_val, range1_limit, range2_limit, plot_max_val]); step_colors = range_colors if benchmark_val >= 0 else range_colors[::-1]; steps = [{'range': [step_ranges[j], step_ranges[j+1]], 'color': step_colors[j]} for j in range(3)]

           # Format title and delta colors based on whether lower is better
           bench_fmt_title = f"{benchmark_val:.1f}%" if is_percent else f"{benchmark_val:.2f}"; title_text = f"<b>{ratio_name}</b><br><i style='font-size:0.8em'>(Bench: {bench_fmt_title} [{bench_src_plot.replace('Fallback ','')}])</i>"
           lower_is_better = ratio_name in ['P/E', 'P/B', 'EV/EBITDA']; delta_decreasing_color = good_color if lower_is_better else bad_color; delta_increasing_color = bad_color if lower_is_better else good_color
           fig_ratios.add_trace(go.Indicator(
                mode = "number+gauge+delta", value = company_val, number={'suffix': "%" if is_percent else "", 'valueformat': ".1f"}, # Format number display
                delta = {'reference': benchmark_val, 'relative': True, 'valueformat':".0%", 'decreasing': {'color': delta_decreasing_color}, 'increasing': {'color': delta_increasing_color}},
                domain = {'row': i, 'column': 0}, title = {'text' : title_text, 'font':{'size':11}},
                gauge = { 'shape': "bullet", 'axis': {'range': [plot_min_val, plot_max_val] if plot_min_val < plot_max_val else [plot_min_val, plot_min_val+1], 'tickformat':'.1f'}, # Use calculated axis range
                    'threshold': { 'line': {'color': "black", 'width': 2}, 'thickness': 0.8, 'value': benchmark_val }, # Show benchmark line
                    'steps': [ s for s in steps if s and isinstance(s.get('range'), list) and len(s['range'])==2 and s['range'][0] < s['range'][1] ], # Use calculated steps, ensure valid
                    'bar': {'color': "#2c3e50", 'thickness': 0.5} }
            ))
       fig_ratios.update_layout(height = max(300, num_ratios * 85), grid = {'rows': num_ratios, 'columns': 1, 'pattern': "independent"}, title_text='Ratio Performance vs. Benchmarks', title_x=0.5, margin=dict(l=160, r=40, t=60, b=30))
       fig_ratios.show()
    else: display(Markdown("*No valid ratios/benchmarks for comparison chart.*"))

    return ratios, ratio_data_for_table

# Execute Ratio Analysis
if 'error' not in final_valuation or final_valuation['error'] is None:
    # ----> Pass flag <----
    ratios_calculated, ratio_table = analyze_ratios(
        stock_data['info'], stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'],
        final_valuation.get('enterprise_value'), current_price, shares_outstanding, currency_symbol,
        user_benchmarks, benchmark_sources, wacc_components, is_indian_stock )
else: display(Markdown("❌ **Skipping Ratio Analysis due to prior errors.**")); ratios_calculated, ratio_table = {}, []

# %% [markdown]
# ## 9. Summary & Visualization Dashboard

# %%
# Dashboard Function Definition (No changes needed here)
def display_summary_dashboard(company_nm, ticker_sym, current_pr, intrinsic_val_per_sh,
                              wacc_val, term_growth, init_growth, growth_src,
                              final_val_dict, hist_df, fcff_res, currency_sym, info_dict, input_src_dict,
                              is_indian): # Added flag

    display(Markdown(f"## Valuation Summary Dashboard: {company_nm} ({ticker_sym})"))
    source_summary = " | ".join(f"{k}:`{v.split('(')[0].strip()}`" for k, v in input_src_dict.items())
    # Display the source of the initial growth rate
    display(Markdown(f"**Input Sources:** {source_summary} | Growth:`{growth_src}`")) # Show full source string

    # Check validity of key inputs for display logic
    valid_intrinsic = intrinsic_val_per_sh is not None and np.isfinite(intrinsic_val_per_sh) # Allow <= 0 intrinsic
    valid_current_pr = current_pr is not None and np.isfinite(current_pr) and current_pr > 0
    valid_wacc = wacc_val is not None and np.isfinite(wacc_val)
    valid_term_g = term_growth is not None and np.isfinite(term_growth)
    valid_init_g = init_growth is not None and np.isfinite(init_growth)

    # --- Indicator Subplots ---
    fig_indicators = make_subplots(rows=1, cols=3, specs=[[{'type': 'indicator'}, {'type': 'indicator'}, {'type': 'indicator'}]],
                                 subplot_titles=("Price vs. Intrinsic", "WACC", "Growth Rates"), horizontal_spacing=0.1, column_widths=[0.4, 0.3, 0.3])

    # Indicator 1: Price vs Intrinsic
    intrinsic_val_num = intrinsic_val_per_sh if valid_intrinsic else np.nan
    # ----> Pass flag <----
    intrinsic_val_fmt = format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian) if valid_intrinsic else "N/A"
    current_pr_num = current_pr if valid_current_pr else 0.0; upside_fmt = ""; delta_config = None
    if valid_intrinsic and valid_current_pr :
        if intrinsic_val_num > 0: upside_pct = (intrinsic_val_num / current_pr_num) - 1; upside_fmt = f" ({upside_pct:+.1%})"
        comparison_text = f"Intrinsic: {intrinsic_val_fmt}{upside_fmt}"
        # Delta prefix needs to adapt - use symbol directly
        delta_config = {'reference': intrinsic_val_num, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym, 'decreasing': {'color': '#2ecc71', 'symbol': '▲ Upside'}, 'increasing': {'color': '#e74c3c', 'symbol': '▼ Downside'}} # Delta vs Intrinsic
    elif valid_intrinsic: comparison_text = f"Intrinsic: {intrinsic_val_fmt} (Current N/A)"
    else: comparison_text = f"Intrinsic: N/A"
    # Number prefix needs to adapt
    fig_indicators.add_trace(go.Indicator(mode = "number+delta" if delta_config else "number", value = current_pr_num if valid_current_pr else 0, number={'prefix': currency_sym + " " if is_indian else currency_sym, 'valueformat': '.2f', 'font':{'size':24}}, title = {"text": f"<span style='font-size:0.8em'>{comparison_text}</span>"}, delta = delta_config, domain = {'row': 0, 'column': 0}), row=1, col=1)

    # Indicator 2: WACC Gauge
    wacc_val_num = wacc_val * 100 if valid_wacc else None # Use None if invalid
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = wacc_val_num, # Pass None value if invalid
        number = {'suffix': "%",
                  'valueformat':'.1f' if valid_wacc else '', # Only format if valid
                  'font':{'size':20}}, # REMOVED INVALID 'text' property here
        gauge = {'axis': {'range': [MIN_WACC*100, MAX_WACC*100]}, 'bar': {'color': "darkblue" if valid_wacc else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc", 'steps' : [ {'range': [MIN_WACC*100, 9], 'color': "#d5f5e3"}, {'range': [9, 14], 'color': "#fdebd0"}, {'range': [14, MAX_WACC*100], 'color': "#fadbd8"}]} if valid_wacc else None, # Conditional gauge
        title = {"text": "<span style='font-size:0.8em'>WACC</span>"}, domain = {'row': 0, 'column': 1}), row=1, col=2)

    # Indicator 3: Growth Gauge
    term_growth_num = term_growth * 100 if valid_term_g else None; init_growth_fmt = format_percent(init_growth, 1) if valid_init_g else "N/A"
    growth_title_text = f"<span style='font-size:0.8em'>Terminal Growth (g)<br>(Initial: {init_growth_fmt})</span>"
    # Dynamic range for growth gauge
    gauge_max_range = max(TERMINAL_GROWTH_RATE_CAP * 100 + 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) + 1)
    # Use the terminal growth floor (0.0%) for the lower gauge range
    gauge_min_range = MIN_GROWTH_FLOOR * 100 - 1
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = term_growth_num, # Pass None value if invalid
         number = {'suffix': "%",
                   'valueformat':'.1f' if valid_term_g else '', # Only format if valid
                   'font':{'size':20}},
         gauge = {'axis': {'range': [gauge_min_range, gauge_max_range], 'tickwidth': 1}, 'bar': {'color': "darkgrey" if valid_term_g else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [{'range': [gauge_min_range, 2.0], 'color': "lightcoral"}, {'range': [2.0, 4.5], 'color': "lightyellow"}, {'range': [4.5, gauge_max_range], 'color': "lightgreen"}],
                 'threshold' : {'line': {'color': "black", 'width': 3}, 'thickness': 0.75, 'value': term_growth_num } if valid_term_g else None } if valid_term_g else None, # Conditional gauge/threshold
         title = {"text": growth_title_text}, domain = {'row': 0, 'column': 2}), row=1, col=3)

    fig_indicators.update_layout(height=220, margin=dict(l=30, r=30, t=50, b=20)); fig_indicators.show()

    # --- Historical Price Chart ---
    if not hist_df.empty:
        # ----> Use currency symbol in axis title <----
        fig_hist = px.line(hist_df, x=hist_df.index, y='Close', title=f'{company_nm} Price History vs. DCF'); fig_hist.update_layout(xaxis_title='Date', yaxis_title=f'Price ({currency_sym})', height=400, legend_title_text='Legend'); fig_hist.update_xaxes(rangeslider_visible=False)
        # ----> Pass flag to format_currency in annotations <----
        if valid_intrinsic: fig_hist.add_hline(y=intrinsic_val_num, line_dash="dash", line_color="#e74c3c", opacity=0.8, annotation_text=f"DCF Intrinsic: {format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian)}", annotation_position="bottom right", annotation_font_color="#e74c3c")
        if valid_current_pr: fig_hist.add_hline(y=current_pr_num, line_dash="dot", line_color="#3498db", opacity=0.8, annotation_text=f"Current: {format_currency(current_pr_num, currency_sym, is_indian=is_indian)}", annotation_position="top left", annotation_font_color="#3498db")
        fig_hist.show()
    else: display(Markdown("*Historical price data unavailable.*"))

    # --- DCF EV Breakdown Pie Chart ---
    pv_forecast_sum = fcff_res.get('cumulative_pv_fcff', np.nan) if fcff_res else np.nan; pv_tv = fcff_res.get('pv_terminal_value', np.nan) if fcff_res else np.nan
    valid_pvs = np.isfinite([pv_forecast_sum, pv_tv]); total_ev = (pv_forecast_sum + pv_tv) if all(valid_pvs) else np.nan
    if np.isfinite(total_ev) and total_ev != 0 : # EV Breakdown Pie Chart
        labels_pie = []; values_pie = []; pull_pie = []; colors = ['#4e79a7','#f28e2b']
        # Only add components if they are non-zero (handle potential zero TV or PV FCFFs)
        if abs(pv_forecast_sum) > 1e-6: labels_pie.append(f'PV({FORECAST_YEARS}yr FCFFs)'); values_pie.append(pv_forecast_sum); pull_pie.append(0) # Check abs value
        if abs(pv_tv) > 1e-6: labels_pie.append('PV(TV)'); values_pie.append(pv_tv); pull_pie.append(0.05 if len(labels_pie) > 1 else 0) # Check abs value
        if labels_pie: # Only plot if there's something to plot
            # Format hover template based on currency
            hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>" # Default
            if is_indian:
                # Can't easily do dynamic Lakh/Crore in hovertemplate, use standard Rupee format
                 hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
            # ----> Pass flag <----
            fig_dcf_pie = go.Figure(data=[go.Pie(labels=labels_pie, values=values_pie, hole=.4, pull=pull_pie, marker_colors=colors[:len(labels_pie)], hovertemplate=hover_format, textinfo='percent+label', insidetextorientation='radial', sort=False)])
            fig_dcf_pie.update_layout( title_text=f"DCF EV Breakdown (Total EV: {format_currency(total_ev, currency_sym, is_indian=is_indian)})", title_x=0.5, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), height=350, margin=dict(t=60, b=50, l=20, r=20) )
            fig_dcf_pie.show()
        else: display(Markdown(f"*DCF components zero/negligible. Cannot generate EV Breakdown chart.*"))
    else: display(Markdown(f"*EV calculation failed ({format_currency(total_ev, currency_sym, is_indian=is_indian)}). Cannot generate EV Breakdown chart.*")) # Added value for debug

    # --- FCFF Projection Bar Chart ---
    if fcff_res and fcff_res.get('projected_fcffs'):
        years = [f'Y{i+1}' for i in range(len(fcff_res['projected_fcffs']))]; pv_list_clean = [p if np.isfinite(p) else 0 for p in fcff_res.get('pv_fcffs', [])]; fcff_list_clean = [f if np.isfinite(f) else 0 for f in fcff_res.get('projected_fcffs', [])]
        if len(pv_list_clean) == len(years) and len(fcff_list_clean) == len(years):
            df_fcff_plot = pd.DataFrame({'Year': years, 'Forecasted FCFF': fcff_list_clean, 'PV of FCFF': pv_list_clean}); df_fcff_plot['Forecasted FCFF'] = pd.to_numeric(df_fcff_plot['Forecasted FCFF']); df_fcff_plot['PV of FCFF'] = pd.to_numeric(df_fcff_plot['PV of FCFF'])
            df_fcff_melt = df_fcff_plot.melt(id_vars='Year', var_name='Cash Flow Type', value_name='Value')
            # ----> Use currency symbol in axis label <----
            fig_fcff_bar = px.bar(df_fcff_melt, x='Year', y='Value', color='Cash Flow Type', title='FCFF Forecast & Present Values', labels={'Value':f'Value ({currency_sym})'}, barmode='group', height=350, color_discrete_map={'Forecasted FCFF':'#5dade2', 'PV of FCFF':'#af7ac5'})
            # Format hover template based on currency
            hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>" # Default
            if is_indian:
                 hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>" # Cannot easily do Lakh/Crore here
            fig_fcff_bar.update_traces(hovertemplate=hover_bar_format); fig_fcff_bar.update_layout(margin=dict(t=60, b=30, l=50, r=30)); fig_fcff_bar.show()
        else: display(Markdown("*FCFF projection data mismatch/invalid. Cannot plot.*"))
    else: display(Markdown("*FCFF projection data unavailable.*"))

    # --- Valuation Conclusion ---
    display(Markdown("### Valuation Conclusion")); intrinsic_val = final_val_dict.get('intrinsic_value_per_share') if final_val_dict else np.nan
    current_pr_num = current_pr if valid_current_pr else np.nan; conclusion = ""
    if final_val_dict is None or final_val_dict.get('error'): conclusion = f"<span style='color:red'>❌ Valuation Error: {final_val_dict.get('error', '?')}</span>"
    # ----> Pass flag to format_currency calls <----
    elif not np.isfinite(intrinsic_val) or intrinsic_val <= 0: conclusion = f"⚠️ Valuation resulted in invalid/non-positive Intrinsic Value ({format_currency(intrinsic_val, currency_sym, is_indian=is_indian)})."
    elif not np.isfinite(current_pr_num): conclusion = f"✅ Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}. Cannot compare (Current Price N/A)."
    else:
         diff_pct = (intrinsic_val / current_pr_num) - 1
         current_fmt = format_currency(current_pr_num, currency_sym, is_indian=is_indian)
         intrinsic_fmt = format_currency(intrinsic_val, currency_sym, is_indian=is_indian)
         if abs(diff_pct) < 0.10: conclusion = f"<span style='color:grey'>~ Fairly Valued:</span> Price ({current_fmt}) vs DCF ({intrinsic_fmt}). Diff: {diff_pct:+.1%}"
         elif diff_pct < 0: overval_pct = (current_pr_num / intrinsic_val) - 1; conclusion = f"<span style='color:#e74c3c'>🔻 Potentially Overvalued:</span> Price ({current_fmt}) is **{overval_pct:+.1%}** above DCF ({intrinsic_fmt})."
         else: conclusion = f"<span style='color:#2ecc71'>✅ Potentially Undervalued:</span> Price ({current_fmt}) suggests **{diff_pct:+.1%}** upside to DCF ({intrinsic_fmt})."
    display(Markdown(f"> **{conclusion}**"))

# Execute Summary Dashboard Generation
if 'error' not in final_valuation or final_valuation['error'] is None:
    # ----> Pass flag <----
    display_summary_dashboard(
        company_name, ticker_symbol, current_price, final_valuation.get('intrinsic_value_per_share'),
        discount_rate, fcff_results.get('terminal_growth_rate'), fcff_results.get('initial_growth_rate'),
        fcff_results.get('growth_source', 'N/A'), final_valuation, stock_data['history'], fcff_results,
        currency_symbol, stock_data['info'], input_sources, is_indian_stock )
else: display(Markdown("❌ **Skipping Summary Dashboard due to prior errors.**"))


# %% [markdown]
# ## 10. Assumptions & Disclaimer (Reflects Dynamic/User Inputs & New Initial Growth Logic)

# %%
# Dynamically generate assumption strings based on final values (handle potential missing dicts)
# Get the actual source string for Rf (which now includes the default value if used)
rf_source_disp = input_sources.get('Rf', '?') # Already captures default info if applicable
rm_source_disp = input_sources.get('Rm','?')
beta_source_disp = input_sources.get('Beta','?')
c_spread_source_disp = input_sources.get('Credit Spread','N/A')
bench_source_disp = input_sources.get('Benchmarks','?')
growth_source_disp = fcff_results.get('growth_source', 'N/A') if fcff_results and isinstance(fcff_results, dict) else 'N/A'
tax_rate_source_disp = wacc_components.get('tax_rate_source', 'N/A') if wacc_components and isinstance(wacc_components, dict) else 'N/A'
equity_source_disp = wacc_components.get('equity_source','N/A') if wacc_components and isinstance(wacc_components, dict) else 'N/A'
weights_source_disp = wacc_components.get('weights_source','N/A') if wacc_components and isinstance(wacc_components, dict) else 'N/A'
terminal_growth_final = fcff_results.get('terminal_growth_rate', np.nan) if fcff_results and isinstance(fcff_results, dict) else np.nan
initial_growth_final = fcff_results.get('initial_growth_rate', np.nan) if fcff_results and isinstance(fcff_results, dict) else np.nan


# %%
# ----> UPDATED TEXT TO REFLECT **NEW** INITIAL GROWTH LOGIC <----
display(Markdown(f"""
**Key Assumptions & Input Sources:**
*   **Financial Data:** Relies on data provided by Yahoo Finance (`yfinance`). Accuracy subject to provider. Missing data uses fallbacks or causes errors.
*   **Risk-Free Rate (Rf):** Attempted dynamic fetch, **confirmed/overridden by user.** (`Source: {rf_source_disp}` - Default suggestion: `{DEFAULT_RF_RATE_SUGGESTION:.1f}%`)
*   **Expected Market Return (Rm):** Based on **user input.** (`Source: {rm_source_disp}`)
*   **Market Risk Premium (MRP):** Derived from user `Rm` and final `Rf`.
*   **Beta (β):** Dynamically fetched if valid, otherwise **required user input.** (`Source: {beta_source_disp}`)
*   **Cost of Debt (Kd):** Implied if possible & within bounds (`{format_percent(MIN_COST_OF_DEBT)}` - `{format_percent(MAX_COST_OF_DEBT)}`). Otherwise, defaults to `Rf + Credit Spread`.
*   **Credit Spread:** Used only if default Kd triggered. Based on **user input** (default suggestion: `{format_percent(DEFAULT_CREDIT_SPREAD)}`). (`Source: {c_spread_source_disp}`)
*   **Growth Rates:**
    *   **Initial Growth:** `{format_percent(initial_growth_final)}` (`Source: {growth_source_disp}`):
        1.  Calculated using **Last 3 Years of Revenue CAGR.**
        2.  If revenue data insufficient (need 4 years), defaults to **`{format_percent(DEFAULT_INITIAL_GROWTH_RATE)}`**.
        3.  Result bounded between **`{format_percent(MIN_INITIAL_GROWTH_RATE)}`** (floor) and **`{format_percent(MAX_INITIAL_GROWTH_RATE)}`** (ceiling).
    *   **Terminal Growth (g):** `{format_percent(terminal_growth_final)}` = min(GDP cap=`{format_percent(TERMINAL_GROWTH_RATE_CAP)}`, WACC*`0.67`, Initial Growth, Rf*`0.95`), floored at `{format_percent(MIN_GROWTH_FLOOR)}`, ensured `< WACC`.
*   **Tax Rate:** Effective rate `{tax_rate_source_disp}` (bounded `{format_percent(MIN_TAX_RATE)}` - `{format_percent(MAX_TAX_RATE)}`) or default (`{format_percent(DEFAULT_TAX_RATE)}`).
*   **Capital Structure:** Assumed stable using `{equity_source_disp}` and calculated debt. (`Weights Source: {weights_source_disp}`)
*   **Industry Benchmarks:** Uses **user input** or **generic fallbacks** if input left blank. (`Source: {bench_source_disp}`)
*   **Currency Formatting:** Uses ₹/Lakh/Crore for `.NS` tickers, standard K/M/B/T otherwise.

**Limitations:**
*   **Model Sensitivity:** DCF highly sensitive to inputs. **Sensitivity analysis (not included) crucial.**
*   **Input Quality:** Validity depends on user inputs and appropriateness of fallbacks.
*   **Historical Data Availability:** Financial statement data (needed for Revenue CAGR) is often limited to ~4 years via `yfinance`, potentially requiring use of the default initial growth rate if insufficient history exists.
*   **Revenue Volatility:** CAGR can be skewed by outliers in start/end years.
*   **Dynamic Data Reliability:** Depends on `yfinance` accuracy/availability. Rf proxies need verification.
*   **Simplifications:** Linear decay, constant rates/structure. Ignores qualitative factors, macro shocks, sentiment.
*   **Fallback Accuracy:** Generic benchmarks may not be representative. Default growth rate is a simplification.
*   **Data Completeness:** Performance degrades with missing data. *Cross-verify key data.*

**Disclaimer:**
*For **educational and illustrative purposes only**. **NOT investment advice**. Models have limitations. Users MUST conduct own due diligence and consult a qualified financial advisor before investing. Creators NOT liable for actions based on this report.*
"""))










# %%
````

````python
# %% [markdown]
# # Dynamic Stock Financial Analysis & DCF Valuation
# **(Interactive Report - Dynamic/User Input Driven Assumptions - V9 - Simplified Initial Growth)**
#
# **Analyst:** Simulated AI Research Analyst
#
# **Objective:** To perform a comprehensive financial analysis and estimate the intrinsic value per share of a publicly traded company using the Discounted Cash Flow (DCF) methodology. Prioritizes dynamic fetching, falls back to user input, and uses configured defaults as a last resort. Includes fallback benchmarks for ratios and robust error handling/display. Adapts currency formatting for Indian stocks. **Initial growth uses 3yr Revenue Avg Growth (handles negatives), with bounds & default.**
#
# **Methodology:**
# 1.  **User Input & Dynamic Fetching:** Get ticker. Attempt dynamic Rf, require user confirmation/input for Rf, Rm, Beta (if needed), Credit Spread (if needed), and Ratio Benchmarks (with fallbacks). Detects Indian stocks for currency formatting.
# 2.  **Data Acquisition:** Use `yfinance` for statements, info, market data. Fetch extended price history.
# 3.  **WACC Calculation:** Determine WACC using fetched/user inputs via CAPM. Handle errors.
# 4.  **FCFF Calculation & Projection:** Calculate historical FCFF. **Estimate initial growth using 3-year average Revenue growth (modified to handle negatives), bounded between 1.1% and 15%, defaulting to 6.5% if data unavailable.** Project FCFF forward using linear decay to terminal growth. Handle errors.
# 5.  **Terminal Value:** Estimate value beyond the explicit forecast period.
# 6.  **Discounting & Valuation:** Discount cash flows and terminal value. Handle errors.
# 7.  **Ratio Analysis:** Calculate ratios and compare to user-input or fallback default benchmarks. Handles non-finite values.
# 8.  **Interactive Visualization:** Present findings using Plotly, including dynamic gauge ranges, robust display logic, and ₹/Lakh/Crore formatting for Indian stocks.

# %% [markdown]
# ## 1. Setup and Libraries

# %%
# Standard Libraries
import warnings
import sys
import numpy as np
import pandas as pd
import datetime
import math # For isnan, isinf checks

# Data Fetching
import yfinance as yf

# Visualization (Using Plotly)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

# Display Utilities
from IPython.display import display, Markdown, HTML

# Settings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pio.templates.default = "plotly_white" # Cleaner Plotly template

# %% [markdown]
# ## 2. Configuration and Helper Functions (Default Growth Updated)

# %%
# --- Model Configuration ---
DEFAULT_TICKER = "WIPRO.NS" # Default if user input fails
FORECAST_YEARS = 5
# Terminal Growth Rate CAP (Also acts as a default ceiling)
TERMINAL_GROWTH_RATE_CAP = 0.05 # Default/Cap set to 5%
# ----> DEFAULT INITIAL GROWTH (Used only if 3yr Revenue Growth calc fails) <----
DEFAULT_GROWTH_RATE = 0.065  # Default set to 6.5%
# Default Risk-Free Rate suggestion (used if dynamic fetch fails)
DEFAULT_RF_RATE_SUGGESTION = 6.0 # Default set to 6.0% (as percent)

DEFAULT_CURRENCY_SYMBOL = "₹" # Changed default to Rupee, will be overridden by yfinance if possible
DEFAULT_TAX_RATE = 0.25     # General fallback tax rate if calculation fails
DEFAULT_CREDIT_SPREAD = 0.03 # Fallback spread over Rf for Cost of Debt
DEFAULT_MARKET_RISK_PREMIUM = 0.08 # Used only to suggest Rm to user

# --- Generic Fallback Benchmark Ratios (Used ONLY if User leaves input blank) ---
DEFAULT_BENCHMARK_RATIOS = {
    'P/E': 20.0,
    'P/B': 2.5,
    'EV/EBITDA': 12.0,
    'ROE': 15.0, # As percent value (e.g., 15 for 15%)
    'ROA': 8.0,  # As percent value
    'ROCE': 12.0,# As percent value
    'Asset Turnover': 0.8,
    'Current Ratio': 1.5,
    'Quick Ratio': 1.0
}

# --- Model Stability Bounds --- (Not primary financial assumptions, but guardrails)
MIN_WACC = 0.06
MAX_WACC = 0.25
MIN_TAX_RATE = 0.10
MAX_TAX_RATE = 0.40
MIN_GROWTH = 0.00 # Minimum growth floor (for terminal growth and projection steps)
# ----> NEW Bounds for CALCULATED INITIAL GROWTH RATE <----
MIN_INITIAL_GROWTH_FLOOR = 0.011 # 1.1%
MAX_INITIAL_GROWTH_CEILING = 0.15  # 15.0%
# (Old MAX_INITIAL_GROWTH is replaced by MAX_INITIAL_GROWTH_CEILING)

MIN_COST_OF_DEBT = 0.03
MAX_COST_OF_DEBT = 0.15

# --- Ratio Keys Needed for Benchmarking ---
REQUIRED_BENCHMARK_RATIOS = list(DEFAULT_BENCHMARK_RATIOS.keys()) # Use keys from default dict

# --- Helper Functions ---
def print_keys(statement, name):
    """Prints available keys in a yfinance statement for debugging"""
    print(f"\n--- Available Keys in {name} ---")
    if isinstance(statement, pd.DataFrame):
        valid_indices = [idx for idx in statement.index if idx is not None]
        print(", ".join(map(str, valid_indices)))
    elif isinstance(statement, dict):
        print(", ".join(statement.keys()))
    print("------------------------------------\n")

def safe_get(data, key_list, statement_name="Statement", default=0.0):
    """
    Safely retrieves a numeric value from yfinance data (Series or Dict).
    Tries multiple keys, returns default if not found/invalid. Handles NaN/inf better.
    """
    if not isinstance(key_list, list):
        key_list = [key_list]

    value = None
    for key in key_list:
      current_value = None
      raw_val = None
      if isinstance(data, pd.Series):
          if key in data.index:
              try:
                  raw_val = data.loc[key]
                  # Check if raw_val is already numeric and finite before converting
                  if isinstance(raw_val, (int, float)) and np.isfinite(raw_val):
                      current_value = float(raw_val)
                  # Handle Series case (sometimes yfinance returns single value Series)
                  elif isinstance(raw_val, pd.Series):
                       if not raw_val.empty and pd.notna(raw_val.iloc[0]) and np.isfinite(raw_val.iloc[0]):
                            current_value = float(raw_val.iloc[0])
                  # Handle string conversion only if necessary and check finiteness
                  elif pd.notna(raw_val):
                      try_val = float(raw_val)
                      if np.isfinite(try_val):
                          current_value = try_val
              except (KeyError, TypeError, IndexError, ValueError):
                   pass # Ignore errors for this specific key/conversion
      elif isinstance(data, dict):
          raw_val = data.get(key)
          if raw_val is not None:
              try:
                  if isinstance(raw_val, str) and raw_val.lower() in ['n/a', 'none', '']:
                      pass # Treat as invalid
                  else:
                      temp_val = float(raw_val)
                      if np.isfinite(temp_val): # Check for NaN/inf after conversion
                           current_value = temp_val
              except (TypeError, ValueError):
                   pass

      if current_value is not None:
            value = current_value
            break # Found a valid value

    if value is None:
        value = default # Use the provided default if no valid value found
    return value

def format_currency(value, symbol="₹", precision=2, is_indian=False):
    """
    Formats number as currency string, handles None/NaN/inf.
    Uses Lakh/Crore for Indian stocks if is_indian=True.
    """
    if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
    if abs(value) > 1e18: return "Inf" # Handle extremely large numbers as Inf

    sign = "-" if value < 0 else ""
    abs_val = abs(value)

    if is_indian:
        # Use ₹ symbol regardless of fetched symbol for consistency
        symbol = "₹"
        if abs_val >= 1e7:  # 1 Crore (10 million)
            formatted_val = f"{abs_val / 1e7:.{precision}f} Cr"
        elif abs_val >= 1e5:  # 1 Lakh (100 thousand)
            formatted_val = f"{abs_val / 1e5:.{precision}f} Lakh"
        else:
            formatted_val = f"{abs_val:,.{precision}f}" # Add comma separators for smaller Rupee values
        return f"{sign}{symbol} {formatted_val}" # Add space after symbol
    else:
        # Standard international formatting (K, M, B, T)
        if abs_val >= 1e12: formatted_val = f"{abs_val / 1e12:.{precision}f}T"
        elif abs_val >= 1e9: formatted_val = f"{abs_val / 1e9:.{precision}f}B"
        elif abs_val >= 1e6: formatted_val = f"{abs_val / 1e6:.{precision}f}M"
        elif abs_val >= 1e3: formatted_val = f"{abs_val / 1e3:.{precision}f}K"
        else: formatted_val = f"{abs_val:.{precision}f}"
        return f"{sign}{symbol}{formatted_val}" # Standard symbol prefix

def format_percent(value, precision=1):
     """Formats number as percentage string, handles None/NaN/inf."""
     if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
     if abs(value) > 1e5: return "Inf" # Handle extreme % values

     return f"{value * 100:.{precision}f}%"

def get_validated_float_input(prompt_message, default_value=None, allow_blank=False, min_val=None, max_val=None):
    """Gets and validates user input for a float value, handles defaults, blanks, and bounds."""
    while True:
        prompt_with_default = prompt_message
        if default_value is not None:
            prompt_with_default += f" [Default: {default_value}]: "
        elif allow_blank:
             prompt_with_default += f" [Leave blank to skip/use fallback]: "
        else:
             prompt_with_default += ": "

        try:
            user_input_str = input(prompt_with_default).strip()

            if not user_input_str:
                if allow_blank: return None
                if default_value is not None:
                    try: return float(default_value)
                    except ValueError: raise ValueError(f"Internal Error: Invalid default value '{default_value}'.")
                raise ValueError("Input cannot be empty.")

            if user_input_str.endswith('%'): # Allow user to add % sign optionally
                user_input_str = user_input_str[:-1]

            value = float(user_input_str)

            if min_val is not None and value < min_val:
                 raise ValueError(f"Value must be >= {min_val}.")
            if max_val is not None and value > max_val:
                 raise ValueError(f"Value must be <= {max_val}.")

            return value
        except ValueError as e:
            print(f"❌ Invalid input: {e}. Please enter a valid number.")

def try_get_dynamic_rf(rf_ticker_symbol="^TNX"):
    """Attempts to fetch a dynamic risk-free rate (e.g., US 10Y Treasury)."""
    print(f"Attempting to fetch dynamic Risk-Free Rate ({rf_ticker_symbol})...")
    try:
        rf_ticker = yf.Ticker(rf_ticker_symbol)
        hist = rf_ticker.history(period="5d")
        if not hist.empty and 'Close' in hist.columns:
            last_close = hist['Close'].iloc[-1]
            if pd.notna(last_close):
                dynamic_rf = last_close / 100.0 # Assume Yahoo gives yield as %
                if 0.001 < dynamic_rf < 0.20: # Check bounds 0.1% to 20%
                    print(f"✅ Dynamically fetched Rf ({rf_ticker_symbol}): {format_percent(dynamic_rf)}")
                    return dynamic_rf
                else:
                     print(f"⚠️ Fetched Rf value ({format_percent(dynamic_rf)}) out of typical range. Ignoring.")
                     return None
            else:
                 print(f"⚠️ Fetched Rf data was invalid (NaN) for {rf_ticker_symbol}.")
                 return None
        else:
             print(f"⚠️ Could not fetch recent data for {rf_ticker_symbol}.")
             return None
    except Exception as e:
        print(f"❌ Error fetching dynamic Rf ({rf_ticker_symbol}): {e}")
        return None

# --- Growth Calculation Helpers ---
def calculate_average_growth(values_list):
    """
    Calculates average growth rate from a list of historical values (newest first).
    Handles NaNs, skips growth from zero/negative bases. Returns NaN if no valid growth periods.
    """
    growth_rates = []
    if len(values_list) < 2:
        return np.nan # Need at least two points for one growth rate

    for i in range(len(values_list) - 1):
        val_t = values_list[i]       # More recent value (e.g., T-0)
        val_t_minus_1 = values_list[i+1] # Older value (e.g., T-1)

        if pd.notna(val_t) and pd.notna(val_t_minus_1) and math.isfinite(val_t) and math.isfinite(val_t_minus_1):
            # Check if base year is non-zero (positive OR negative)
            if abs(val_t_minus_1) > 1e-6: # Use a small threshold instead of exact zero
                # Calculate growth rate, handling signs correctly
                g = (val_t / val_t_minus_1) - 1 if val_t_minus_1 > 0 else (val_t / val_t_minus_1) - 1 if val_t_minus_1 < 0 else 0.0
                # Basic check for extreme growth rates (e.g., >1000% or <-100%) - often data errors
                if -10.0 < g < 10.0:
                     growth_rates.append(g)
                # else: print(f"   - Skipping growth calc period {i+1}-{i+2}: Extreme growth {g:.1%}. Values: {val_t_minus_1} -> {val_t}")
            # else: # Optional: log skipped period
            #    print(f"   - Skipping growth calc period {i+1}-{i+2}: Near-zero base value {val_t_minus_1}")
        # else: # Optional: log skipped period
        #    print(f"   - Skipping growth calc period {i+1}-{i+2}: NaN/inf values ({val_t}, {val_t_minus_1})")


    if not growth_rates:
        return np.nan
    else:
        return np.mean(growth_rates)

# calculate_cagr is kept for potential future use or reference, but not used for initial growth
def calculate_cagr(start_val, end_val, num_years):
    """Calculates Compound Annual Growth Rate. Returns NaN for non-positive start_val."""
    if pd.isna(start_val) or pd.isna(end_val) or not math.isfinite(start_val) or not math.isfinite(end_val) or num_years <= 0:
        return np.nan
    if start_val <= 1e-9: # Check for non-positive start value
        return np.nan

    # Store original sign of end value before adjustments
    original_end_val_sign = np.sign(end_val)

    # Adjust end_val if non-positive to allow calculation
    if end_val <= 1e-9:
        end_val = 1e-9 # Use tiny positive for calculation

    try:
        # Calculate CAGR based on absolute ratio
        cagr = (abs(end_val / start_val)) ** (1 / num_years) - 1

        # If the original end value was negative, reflect this as a negative CAGR magnitude
        # A true CAGR isn't really defined for negative end values, but this shows the direction.
        if original_end_val_sign < 0:
             # Make it significantly negative to indicate endpoint negativity, more intuitively than just negating cagr
             cagr = -((abs(end_val / start_val)) ** (1 / num_years)) - 1

    except (ValueError, OverflowError, ZeroDivisionError):
        cagr = np.nan
    return cagr
# --- End Growth Calculation Helpers ---

# %% [markdown]
# ## 3. User Input & Dynamic Fetching: Ticker, Rf, Rm, Benchmarks, Beta, Spread

# %%
input_sources = {}

# --- Get Ticker Symbol ---
user_ticker = input(f"Enter stock ticker (e.g., MSFT, AAPL, RELIANCE.NS) [Default: {DEFAULT_TICKER}]: ")
if not user_ticker: user_ticker = DEFAULT_TICKER
ticker_symbol = user_ticker.strip().upper()
display(Markdown(f"**Analyzing Ticker:** `{ticker_symbol}`"))

# ----> DETECT IF INDIAN STOCK for currency formatting <----
is_indian_stock = ticker_symbol.endswith(".NS")
if is_indian_stock:
    print("Info: Detected Indian stock (.NS). Using ₹/Lakh/Crore formatting.")

# --- Determine RF Ticker based on Stock Ticker ---
rf_proxy_ticker = "^TNX" # Default US 10Y
if is_indian_stock: rf_proxy_ticker = "^NS10"; print(f"Info: Using Indian Rf proxy: {rf_proxy_ticker}")
elif ticker_symbol.endswith(".L"): rf_proxy_ticker = "^FTLC"; print(f"Info: Using UK Rf proxy: {rf_proxy_ticker}")
# Add more country/exchange specific Rf proxies if needed

# --- Get Risk-Free Rate (Rf) ---
print("\n--- CAPM Inputs ---")
dynamic_rf_rate = try_get_dynamic_rf(rf_proxy_ticker)
rf_prompt = f"Enter Risk-Free Rate (Rf) as percentage (e.g., '4.5' for 4.5%). Fetched {rf_proxy_ticker}?"
# ----> Use NEW default suggestion if dynamic fetch fails <----
suggested_rf_val = dynamic_rf_rate * 100 if dynamic_rf_rate is not None else DEFAULT_RF_RATE_SUGGESTION

risk_free_rate_input_pct = get_validated_float_input(
    rf_prompt, default_value=f"{suggested_rf_val:.2f}", allow_blank=False, min_val=0.0, max_val=20.0
)
risk_free_rate_input = risk_free_rate_input_pct / 100.0

input_sources['Rf'] = 'Dynamic Fetch (User Confirmed/Adjusted)' if dynamic_rf_rate is not None else f'User Input (Default: {DEFAULT_RF_RATE_SUGGESTION:.1f}%)' # Updated source string slightly
display(Markdown(f"**Using Risk-Free Rate (Rf):** {format_percent(risk_free_rate_input)} (`Source: {input_sources['Rf']}`)"))

# --- Get Expected Market Return (Rm) ---
suggested_mrp = DEFAULT_MARKET_RISK_PREMIUM
suggested_rm = risk_free_rate_input + suggested_mrp
rm_prompt = f"Enter Expected Market Return (Rm) as percentage (e.g., '10'). Suggestion based on Rf + {format_percent(suggested_mrp)} MRP is {format_percent(suggested_rm)}"

while True:
    market_return_input_pct = get_validated_float_input(
        rm_prompt, default_value=f"{suggested_rm * 100:.1f}", allow_blank=False, min_val=-10.0, max_val=35.0
    )
    market_return_input = market_return_input_pct / 100.0
    if market_return_input >= risk_free_rate_input: break
    else: print(f"❌ Error: Rm ({format_percent(market_return_input)}) must be >= Rf ({format_percent(risk_free_rate_input)}). Re-enter.")

market_risk_premium_input = market_return_input - risk_free_rate_input
input_sources['Rm'] = 'User Input'
input_sources['MRP'] = 'Derived'
display(Markdown(f"**Using Expected Market Return (Rm):** {format_percent(market_return_input)} (`Source: {input_sources['Rm']}`)"))
display(Markdown(f"**↳ Implied Market Risk Premium (MRP):** {format_percent(market_risk_premium_input)} (`Source: {input_sources['MRP']}`)"))

# --- Beta (Placeholder) ---
beta_input = None; input_sources['Beta'] = 'Pending'
# --- Credit Spread (Placeholder) ---
user_credit_spread = None; input_sources['Credit Spread'] = 'Pending'

# --- Get Industry Benchmark Ratios (with fallback defaults) ---
print("\n--- Industry Ratio Benchmarks ---")
display(Markdown("Enter benchmarks or leave blank for generic fallbacks. *(Find industry averages online)*"))
user_benchmarks = {}; benchmark_sources = {}
print("Provide benchmarks (e.g., ROE: enter '15' for 15%; P/E: enter '20'):")
for ratio_key in REQUIRED_BENCHMARK_RATIOS:
    default_val = DEFAULT_BENCHMARK_RATIOS.get(ratio_key, None)
    prompt = f"  Enter benchmark for {ratio_key}"
    is_percent_type = ratio_key in ['ROE', 'ROA', 'ROCE']
    min_b, max_b = (-50, 100) if is_percent_type else (-100, 500)

    bench_val_input = get_validated_float_input(
        prompt, default_value=f"{default_val:.2f}" if default_val is not None else None,
        allow_blank=True, min_val=min_b, max_val=max_b
    )

    if bench_val_input is not None:
        user_benchmarks[ratio_key] = bench_val_input
        benchmark_sources[ratio_key] = 'User Input'
        display_val = f"{bench_val_input:.1f}%" if is_percent_type else f"{bench_val_input:.2f}"
        display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: User`)"))
    elif default_val is not None:
         user_benchmarks[ratio_key] = default_val # Store the default value if user skipped
         benchmark_sources[ratio_key] = 'Fallback Default'
         display_val = f"{default_val:.1f}%" if is_percent_type else f"{default_val:.2f}"
         display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: Default`)"))
    else:
        benchmark_sources[ratio_key] = 'Not Available'
        display(Markdown(f"  - **{ratio_key} Benchmark:** Not Provided / No Default"))
input_sources['Benchmarks'] = 'User Input or Default'

# %% [markdown]
# ## 4. Data Acquisition (Fetching Extended History)

# %%
stock_data = {}
try:
    stock = yf.Ticker(ticker_symbol)
    info = stock.info
    if not info or not info.get('regularMarketPrice'):
        print(f"Warning: Limited info for {ticker_symbol}. Trying history...")
        hist_check = stock.history(period="5d");
        if hist_check.empty: raise ValueError(f"Ticker '{ticker_symbol}' invalid/no data.")
        else: print(f"Warning: Proceeding with limited data."); info = {} if not isinstance(info, dict) else info
    stock_data['info'] = info

    # Standard statement fetching (usually 4 years)
    stock_data['financials'] = stock.financials
    stock_data['balance_sheet'] = stock.balance_sheet
    stock_data['cashflow'] = stock.cashflow
    if stock_data['financials'].empty or stock_data['balance_sheet'].empty or stock_data['cashflow'].empty:
       print(f"⚠️ Warning: Empty statements for {ticker_symbol}. DCF unreliable.")

    # Fetch recent history for charts/current price fallback
    stock_data['history'] = stock.history(period="2y")
    # ----> Fetch extended history (still useful for context/reference even if not primary for growth) <----
    stock_data['history_long'] = stock.history(period="11y") # Fetch 11 years for reference
    if stock_data['history_long'].empty:
        print(f"⚠️ Warning: Could not fetch long history (11y) for {ticker_symbol}.")


    company_name = info.get("longName", info.get("shortName", ticker_symbol))
    # Determine currency symbol, prioritizing Rupee for Indian stocks
    currency_code = info.get("currency", DEFAULT_CURRENCY_SYMBOL)
    if not isinstance(currency_code, str) or len(currency_code)>5 : currency_code=DEFAULT_CURRENCY_SYMBOL
    currency_symbol = "₹" if is_indian_stock else info.get("symbol", currency_code) # Prefer symbol if available, else code, force Rupee for Indian

    current_price = safe_get(info, ['currentPrice', 'regularMarketPrice', 'previousClose', 'lastPrice'], 'Info', default=np.nan)
    price_source = "yfinance Info"
    if not np.isfinite(current_price) and not stock_data['history'].empty:
        last_close_hist = stock_data['history']['Close'].iloc[-1]
        if pd.notna(last_close_hist): current_price = last_close_hist; price_source = "yfinance History"
    if not np.isfinite(current_price): print(f"⚠️ Warning: Could not determine current price."); current_price = np.nan

    shares_outstanding = safe_get(info, ['sharesOutstanding'], 'Info', default=0)

    beta_input = safe_get(info, ['beta'], 'Info', default=None)
    if beta_input is not None and np.isfinite(beta_input):
        input_sources['Beta'] = 'Dynamic Fetch'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))
    else:
         print(f"❌ ERROR: Beta not found dynamically or invalid.")
         beta_input = get_validated_float_input(f"Enter Beta for {ticker_symbol}:", allow_blank=False, min_val=-1.0, max_val=4.0)
         info['beta'] = beta_input; input_sources['Beta'] = 'User Input'
         display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))

    market_cap = safe_get(info, ['marketCap'], 'Info', default=0)
    mcap_source = "yfinance Info"
    if market_cap <= 0 and shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Calculated"; print(f"Info: Market Cap calculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}") # Pass flag
    elif market_cap <= 0: mcap_source = "Unavailable"; print(f"⚠️ Warning: Market Cap unavailable."); market_cap = np.nan

    sector = info.get("sector", "N/A"); industry = info.get("industry", "N/A")
    summary = info.get("longBusinessSummary", "No summary available.")

    display(Markdown(f"**Company:** {company_name}")); display(Markdown(f"**Sector:** {sector} | **Industry:** {industry}"))
    display(Markdown(f"**Currency:** {currency_code} ({currency_symbol})")); # Show both code and effective symbol
    # ----> Pass is_indian_stock flag to format_currency calls <----
    display(Markdown(f"**Current Market Price:** {format_currency(current_price, currency_symbol, is_indian=is_indian_stock)} (`Source: {price_source}`)"))
    display(Markdown(f"**Market Cap:** {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)} (`Source: {mcap_source}`)"))

except Exception as e: print(f"❌ CRITICAL ERROR fetching data: {e}"); sys.exit("Cannot proceed.")

if not isinstance(shares_outstanding, (int, float)) or shares_outstanding <= 0: sys.exit("Essential data missing: Shares Outstanding.")
if beta_input is None or not np.isfinite(beta_input): sys.exit("Essential data missing: Beta.")
if not np.isfinite(market_cap): print(f"⚠️ WARNING: Market Cap unavailable. WACC weights less accurate.")

# %% [markdown]
# ## 5. WACC Calculation

# %%
# WACC Function Definition (No changes here)
def calculate_wacc(stock_info_dict, financials_df, balance_sheet_df,
                   user_rf, user_mrp, user_beta, input_sources_dict,
                   currency_sym, is_indian): # Added currency params
    global user_credit_spread
    display(Markdown("### Calculating WACC..."))
    components = {}
    calculation_log = []
    calculation_log.append(f"Input Sources: Rf='{input_sources_dict.get('Rf')}', MRP='{input_sources_dict.get('MRP')}', Beta='{input_sources_dict.get('Beta')}'")

    beta_val = user_beta
    if beta_val <= 0: calculation_log.append(f"⚠️ Warning: Non-positive Beta ({beta_val:.2f}).")
    components['beta'] = beta_val; components['risk_free_rate'] = user_rf; components['market_risk_premium'] = user_mrp
    cost_of_equity = user_rf + beta_val * user_mrp
    components['cost_of_equity'] = cost_of_equity
    calculation_log.append(f"Inputs -> Rf: {format_percent(user_rf)}, Beta: {beta_val:.2f}, MRP: {format_percent(user_mrp)}")
    calculation_log.append(f"↳ Cost of Equity (Ke) [CAPM]: {format_percent(cost_of_equity)}")

    cost_of_debt_pre_tax = np.nan; tax_rate = np.nan; after_tax_cost_of_debt = np.nan
    total_debt = 0.0; debt_calculation_method = "N/A"; tax_rate_source = "N/A"

    if not balance_sheet_df.empty and not financials_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]; latest_fin = financials_df.iloc[:, 0]
        debt_keys = ['Total Debt', 'Long Term Debt', 'Long Term Debt And Capital Lease Obligation', 'Total Debt Non Current']
        long_term_debt = safe_get(latest_bs, debt_keys , 'Balance Sheet', default=0)
        current_debt_keys = ['Current Debt', 'Short Term Debt', 'Current Debt And Capital Lease Obligation', 'Short Long Term Debt', 'Total Debt Current']
        short_term_debt = safe_get(latest_bs, current_debt_keys , 'Balance Sheet', default=0)
        total_debt = long_term_debt + short_term_debt
        components['debt_value_used'] = total_debt
        interest_expense = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], 'Income Statement', default=0))

        implied_rate_calculated = False
        if total_debt > 1000 and interest_expense > 0:
             implied_rate = interest_expense / total_debt; implied_rate_calculated = True
             if MIN_COST_OF_DEBT <= implied_rate <= MAX_COST_OF_DEBT:
                  cost_of_debt_pre_tax = implied_rate; debt_calculation_method = "Implied";
                  # ----> Pass is_indian flag to format_currency calls <----
                  calculation_log.append(f"Interest Exp: {format_currency(interest_expense, symbol=currency_sym, is_indian=is_indian)}, Debt: {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
                  calculation_log.append(f"↳ Implied Cost of Debt (Kd pre-tax): {format_percent(cost_of_debt_pre_tax)}")
             else: calculation_log.append(f"⚠️ Implied Kd ({format_percent(implied_rate)}) out of bounds. Falling back.")
        elif total_debt > 1000: calculation_log.append(f"ℹ️ Found Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}) but no Interest Exp. Falling back.") # Pass flag

        if not debt_calculation_method.startswith("Implied"):
             if total_debt > 1000:
                 debt_calculation_method = "Default (Rf+Spread)"
                 if user_credit_spread is None:
                     spread_prompt = f"Implied Kd failed/NA. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
                     user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
                     user_credit_spread = user_credit_spread_pct / 100.0
                     input_sources_dict['Credit Spread'] = 'User Input (Default Kd)'
                 else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
                 cost_of_debt_pre_tax = user_rf + user_credit_spread
                 calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(cost_of_debt_pre_tax)}")
                 calculation_log.append(f"   (Spread Source: {input_sources_dict['Credit Spread']})")
             else:
                 debt_calculation_method = "Zero Debt"; cost_of_debt_pre_tax = 0.0; input_sources_dict['Credit Spread'] = 'N/A (Zero Debt)';
                 # ----> Pass is_indian flag to format_currency calls <----
                 calculation_log.append(f"ℹ️ Negligible Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}). Kd = 0%.")

        income_tax_expense = safe_get(latest_fin, ['Tax Provision', 'Income Tax Expense'], 'Income Statement', default=0)
        pretax_income = safe_get(latest_fin, ['Pretax Income', 'EBT', 'Earnings Before Tax'], 'Income Statement', default=0)
        if pretax_income <= 0 and income_tax_expense > 0: # Handle negative pretax income if tax was paid
             ebit_val = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=0); interest_exp_for_ebt = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], default=0))
             pretax_income_from_ebit = ebit_val - interest_exp_for_ebt
             if pretax_income_from_ebit > 0: pretax_income = pretax_income_from_ebit # Use calculated if positive
        if pretax_income > 0 and income_tax_expense >= 0:
            calculated_tax_rate = income_tax_expense / pretax_income; tax_rate = min(max(calculated_tax_rate, MIN_TAX_RATE), MAX_TAX_RATE); tax_rate_source = "Derived & Bounded"
            if abs(tax_rate - calculated_tax_rate) > 0.001: tax_rate_source = f"Derived ({format_percent(calculated_tax_rate)}), Bounded"
            # ----> Pass is_indian flag to format_currency calls <----
            calculation_log.append(f"Tax Exp: {format_currency(income_tax_expense, symbol=currency_sym, is_indian=is_indian)}, Pretax Inc: {format_currency(pretax_income, symbol=currency_sym, is_indian=is_indian)}")
            calculation_log.append(f"↳ Effective Tax Rate: {format_percent(tax_rate)} ({tax_rate_source})")
        else: tax_rate = DEFAULT_TAX_RATE; tax_rate_source = "Default"; calculation_log.append(f"⚠️ Cannot derive tax rate. Using default: {format_percent(tax_rate)}")
    else: # Financial statements empty
         calculation_log.append("⚠️ Financial statements empty. Using defaults."); debt_calculation_method = "Default (No Fin.)"; tax_rate_source = "Default (No Fin.)"
         if user_credit_spread is None:
              spread_prompt = f"No financials. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
              user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
              user_credit_spread = user_credit_spread_pct / 100.0; input_sources_dict['Credit Spread'] = 'User Input (No Fin.)'
         else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
         cost_of_debt_pre_tax = user_rf + user_credit_spread; tax_rate = DEFAULT_TAX_RATE; total_debt = 0; components['debt_value_used'] = 0
         calculation_log.append(f"Using Default Kd: {format_percent(cost_of_debt_pre_tax)} (Rf + Spread: {input_sources_dict['Credit Spread']})")
         calculation_log.append(f"Using Default Tax Rate: {format_percent(tax_rate)}")

    components['cost_of_debt'] = cost_of_debt_pre_tax if np.isfinite(cost_of_debt_pre_tax) else 0.0
    components['debt_calc_method'] = debt_calculation_method
    components['tax_rate'] = tax_rate if np.isfinite(tax_rate) else DEFAULT_TAX_RATE
    components['tax_rate_source'] = tax_rate_source
    if np.isfinite(cost_of_debt_pre_tax) and np.isfinite(tax_rate): after_tax_cost_of_debt = cost_of_debt_pre_tax * (1 - tax_rate)
    else: after_tax_cost_of_debt = components['cost_of_debt'] * (1 - DEFAULT_TAX_RATE) # Use default if calculation failed
    components['after_tax_cost_of_debt'] = after_tax_cost_of_debt
    if components['cost_of_debt'] > 0 : calculation_log.append(f"↳ Cost of Debt (After-tax): {format_percent(after_tax_cost_of_debt)}")

    market_cap_val = safe_get(stock_info_dict, ['marketCap'], 'Info', default=np.nan); equity_source = "Market Cap" if np.isfinite(market_cap_val) and market_cap_val > 0 else "N/A"
    equity_value_for_wacc = np.nan
    if np.isfinite(market_cap_val) and market_cap_val > 0: equity_value_for_wacc = market_cap_val
    else:
       calculation_log.append("⚠️ Market Cap unavailable/zero."); equity_source = "Book Equity Fallback"
       if not balance_sheet_df.empty:
           equity_keys = ['Total Stockholder Equity', 'Stockholders Equity', 'Total Equity Gross Minority Interest', 'Common Stock Equity']
           book_equity_val = safe_get(balance_sheet_df.iloc[:, 0], equity_keys, 'Balance Sheet', default=np.nan)
           if np.isfinite(book_equity_val) and book_equity_val > 1.0:
              equity_value_for_wacc = book_equity_val; calculation_log.append(f"Using Book Equity ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}) as proxy.") # Pass flag
           else: equity_source = "Placeholder (Book Invalid)"
       else: equity_source = "Placeholder (No BS)"
       if not np.isfinite(equity_value_for_wacc): equity_value_for_wacc = 1.0; calculation_log.append("⚠️ Market Cap & Book Equity unavailable. Using placeholder Equity Value (1.0). Weights inaccurate.")
    components['equity_value_used'] = equity_value_for_wacc; components['equity_source'] = equity_source

    equity_weight = np.nan; debt_weight = np.nan; weights_source = "N/A"; total_capital = equity_value_for_wacc + total_debt
    if np.isfinite(total_capital) and total_capital > 1.0:
        equity_weight = equity_value_for_wacc / total_capital; debt_weight = total_debt / total_capital; weights_source = "Calculated"
        total_weight = equity_weight + debt_weight # Normalize if calculation slightly off
        if abs(total_weight - 1.0) > 1e-6: equity_weight /= total_weight; debt_weight /= total_weight
    else: calculation_log.append(f"❌ ERROR: Total Capital ({format_currency(total_capital, symbol=currency_sym, is_indian=is_indian)}) invalid. Using default weights (70/30)."); equity_weight = 0.7; debt_weight = 0.3; weights_source = "Default" # Pass flag
    components['equity_weight'] = equity_weight; components['debt_weight'] = debt_weight; components['weights_source'] = weights_source
    # ----> Pass is_indian flag to format_currency calls <----
    calculation_log.append(f"Equity Value Used ({components['equity_source']}): {format_currency(components['equity_value_used'], symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"Debt Value Used ({components['debt_calc_method']}): {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"↳ Weights E/V: {format_percent(equity_weight)}, D/V: {format_percent(debt_weight)} (Source: {weights_source})")

    wacc = np.nan
    if all(np.isfinite(v) for v in [equity_weight, cost_of_equity, debt_weight, after_tax_cost_of_debt]):
        wacc = (equity_weight * cost_of_equity) + (debt_weight * after_tax_cost_of_debt); original_wacc = wacc; wacc = min(max(wacc, MIN_WACC), MAX_WACC)
        bound_note = ""
        if abs(wacc - original_wacc) > 0.0001 : bound_note = f"(Bounded from {format_percent(original_wacc)})"; calculation_log.append(f"⚠️ WACC ({format_percent(original_wacc)}) bounded.")
        calculation_log.append(f"▶️ Calculated WACC: {format_percent(wacc)} {bound_note}")
    else: calculation_log.append(f"❌ ERROR: Cannot calculate final WACC due to invalid components."); wacc = np.nan
    components['final_wacc'] = wacc

    display(Markdown("```\n" + "\n".join(calculation_log) + "\n```"))

    if np.isfinite(wacc) and wacc > 0 and np.isfinite(equity_weight) and np.isfinite(debt_weight):
        equity_contrib = equity_weight * cost_of_equity; debt_contrib = debt_weight * after_tax_cost_of_debt
        # Check if components reasonably sum up (handle potential float issues)
        if np.isfinite(equity_contrib) and np.isfinite(debt_contrib) and abs(equity_contrib + debt_contrib - wacc) < 0.01:
           labels = ['Cost of Equity Contribution', 'Cost of Debt Contribution']; values = [equity_contrib, debt_contrib]
           fig_wacc_donut = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4, marker_colors=px.colors.qualitative.Pastel, hovertemplate='<b>%{label}</b><br>Contribution: %{value:.2%}<extra></extra>', textinfo='percent', textfont_size=12, sort=False)])
           fig_wacc_donut.update_layout(title_text=f'WACC Composition (Total: {format_percent(wacc)})', title_x=0.5, showlegend=True, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), margin=dict(l=20, r=20, t=60, b=50), height=350)
           fig_wacc_donut.show()
        else: display(Markdown(f"*(WACC component contributions invalid/don't sum. Cannot generate chart.)*"))
    elif not np.isfinite(wacc): display(Markdown(f"*(WACC calculation failed. Cannot generate chart)*"))
    else: display(Markdown(f"*(WACC zero/negative or weights invalid. Cannot generate chart)*"))

    return wacc, components

# Execute WACC Calculation
# ----> Pass currency symbol and flag <----
discount_rate, wacc_components = calculate_wacc(
    stock_data['info'], stock_data['financials'], stock_data['balance_sheet'],
    risk_free_rate_input, market_risk_premium_input, beta_input, input_sources,
    currency_symbol, is_indian_stock
)

# %% [markdown]
# ## 6. FCFF Calculation & Projection (Initial Growth: 3yr Revenue Avg, Bounded, Default)

# %%
# FCFF Function Definition (NEW initial growth logic)
def calculate_and_project_fcff(financials_df, balance_sheet_df, cashflow_df,
                               wacc_comps, disc_rate, forecast_yrs,
                               final_user_rf, default_init_growth, # This is 6.5% default
                               currency_sym, is_indian): # Added currency params
    display(Markdown("### Calculating & Projecting FCFF..."))
    projection_log = []
    projection_results = {
        'historical_fcffs': {}, # Store historical FCFFs used for growth calc
        'historical_revenues': {}, # Store historical revenues
        'current_fcff': np.nan, 'projected_fcffs': [], 'growth_rates': [],
        'pv_fcffs': [], 'cumulative_pv_fcff': np.nan, 'terminal_growth_rate': np.nan,
        'terminal_value': np.nan, 'pv_terminal_value': np.nan, 'error': None,
        'initial_growth_rate': np.nan, 'growth_source': 'N/A'
    }

    if financials_df.empty or balance_sheet_df.empty or cashflow_df.empty:
         projection_results['error'] = "Missing Financials/Balance Sheet/Cash Flow."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results
    if not np.isfinite(disc_rate):
         projection_results['error'] = "Invalid WACC provided."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results

    # --- Define Helper to calculate FCFF for a specific year column ---
    # (Same helper as before, ensures consistency)
    def get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, tax_rt):
        if not isinstance(fin_col, pd.Series) or not isinstance(bs_col, pd.Series) or not isinstance(cf_col, pd.Series):
            return np.nan, "Missing Statement Column"

        ebit = safe_get(fin_col, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
        depreciation_amortization = safe_get(cf_col, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
        capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
        capital_expenditures = abs(safe_get(cf_col, capex_keys, default=0))
        wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]
        change_in_wc = safe_get(cf_col, wc_keys, default=None)
        wc_src = "CF"

        if change_in_wc is None or not np.isfinite(change_in_wc):
            wc_src = "BS Calc"
            if isinstance(bs_prev_col, pd.Series): # Need previous year BS
                 curr_assets = safe_get(bs_col, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(bs_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                 prev_assets = safe_get(bs_prev_col, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(bs_prev_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                 if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val):
                     change_in_wc = curr_wc_val - prev_wc_val
                 else: change_in_wc = 0; wc_src = "BS Fail"
            else: change_in_wc = 0; wc_src = "No Prev BS"
        if not np.isfinite(change_in_wc): change_in_wc = 0

        fcff_val = np.nan; nopat_val = np.nan; calc_status = "OK"
        if np.isfinite(ebit) and np.isfinite(tax_rt):
            nopat_val = ebit * (1 - tax_rt)
            if np.isfinite(depreciation_amortization) and np.isfinite(capital_expenditures) and np.isfinite(change_in_wc):
                fcff_val = nopat_val + depreciation_amortization - capital_expenditures - change_in_wc
            else: calc_status = "Invalid Depr/CapEx/WC"
        else: calc_status = "Invalid EBIT/Tax"
        return fcff_val, calc_status
    # --- End FCFF Helper ---

    # --- Calculate Historical FCFF & Revenue (as far back as possible) ---
    max_hist_years = 11 # Allow up to 10 periods of growth (needs 11 data points)
    num_statement_years = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    historical_fcffs_list = [] # Newest first (index 0 = T-0)
    historical_revenues_list = [] # Newest first

    effective_tax_rate = wacc_comps.get('tax_rate', DEFAULT_TAX_RATE) # Use consistent tax rate for historical calc

    projection_log.append("Calculating historical FCFF & Revenue...")
    for i in range(min(num_statement_years, max_hist_years)):
        year_label = f"T-{i}"
        fin_col = financials_df.iloc[:, i]
        bs_col = balance_sheet_df.iloc[:, i]
        cf_col = cashflow_df.iloc[:, i]
        bs_prev_col = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None

        # Calculate FCFF for the year
        fcff_val, status = get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, effective_tax_rate)
        historical_fcffs_list.append(fcff_val if status == "OK" else np.nan) # Store NaN if calc failed

        # Get Revenue for the year
        rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
        revenue_val = safe_get(fin_col, rev_keys, default=np.nan)
        historical_revenues_list.append(revenue_val)

    projection_results['historical_fcffs'] = {f"T-{i}": v for i, v in enumerate(historical_fcffs_list)}
    projection_results['historical_revenues'] = {f"T-{i}": v for i, v in enumerate(historical_revenues_list)}
    projection_results['current_fcff'] = historical_fcffs_list[0] if historical_fcffs_list and np.isfinite(historical_fcffs_list[0]) else np.nan
    fcff_current = projection_results['current_fcff']

    # --- Display Calculation Log for Current FCFF (if calculated) ---
    if np.isfinite(fcff_current):
        latest_fin = financials_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]
        ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
        nopat = ebit * (1-effective_tax_rate) if np.isfinite(ebit) else np.nan
        depreciation_amortization = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
        capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
        capital_expenditures = abs(safe_get(latest_cf, capex_keys, default=0))
        wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]
        change_in_wc = safe_get(latest_cf, wc_keys, default=None)
        wc_source = "Direct from Cash Flow"
        if change_in_wc is None or not np.isfinite(change_in_wc):
            if balance_sheet_df.shape[1] >= 2:
                prev_bs = balance_sheet_df.iloc[:, 1]
                curr_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                prev_assets = safe_get(prev_bs, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(prev_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val): change_in_wc = curr_wc_val - prev_wc_val; wc_source = "Calculated from BS"
                else: change_in_wc = 0; wc_source = "Calc Failed (Invalid BS)"
            else: change_in_wc = 0; wc_source = "Default (No Prev BS)"
        if not np.isfinite(change_in_wc): change_in_wc = 0
        projection_log.append(f"EBIT (T-0): {format_currency(ebit, currency_sym, is_indian=is_indian)}, Tax Rate ({wacc_comps.get('tax_rate_source','?')}) {format_percent(effective_tax_rate)} -> NOPAT: {format_currency(nopat, currency_sym, is_indian=is_indian)}")
        projection_log.append(f"  + Dep&Amort: {format_currency(depreciation_amortization, currency_sym, is_indian=is_indian)}")
        projection_log.append(f"  - CapEx: {format_currency(capital_expenditures, currency_sym, is_indian=is_indian)}")
        projection_log.append(f"  - Change WC ({wc_source}): {format_currency(change_in_wc, currency_sym, is_indian=is_indian)}")
        projection_log.append(f"▶️ Historical FCFF (T-0): {format_currency(fcff_current, currency_sym, is_indian=is_indian)}")
    else:
        projection_log.append("⚠️ Could not calculate current historical FCFF (T-0).")

    if not np.isfinite(fcff_current):
        projection_results['error'] = "Invalid historical FCFF (T-0). Cannot proceed.";
        display(Markdown(f"❌ ERROR: {projection_results['error']}")); display(Markdown("```\n" + "\n".join(projection_log) + "\n```")); return projection_results

    # --- NEW Initial Growth Rate Calculation (3yr Rev Avg -> Default -> Bounds) ---
    projection_log.append("--- Determining Initial Growth Rate ---")
    initial_growth_rate = np.nan # Start with NaN
    growth_source = "N/A"
    found_growth = False

    # 1. Try 3-Year Average Revenue Growth
    projection_log.append("Attempting 3-Year Average Revenue Growth...")
    # Need 4 revenue points for 3 growth periods (T-0/T-1, T-1/T-2, T-2/T-3)
    if len(historical_revenues_list) >= 4:
        rev_slice_3yr = historical_revenues_list[:4] # T-0, T-1, T-2, T-3
        # Check if data in slice is valid enough
        valid_rev_points = sum(1 for r in rev_slice_3yr if pd.notna(r) and math.isfinite(r))
        if valid_rev_points >= 2: # Need at least 2 points to calculate *any* growth
            avg_rev_g_3yr = calculate_average_growth(rev_slice_3yr)
            if pd.notna(avg_rev_g_3yr) and math.isfinite(avg_rev_g_3yr):
                initial_growth_rate = avg_rev_g_3yr # Store the calculated rate
                growth_source = "3yr Avg Revenue Growth"
                projection_log.append(f"  ✅ Success: Calculated {growth_source} = {format_percent(initial_growth_rate)}")
                found_growth = True
            else:
                projection_log.append(f"  - Revenue Growth (3yr): Calculation failed (e.g., invalid intermediate values, zero bases).")
        else:
             projection_log.append(f"  - Revenue Growth (3yr): Insufficient valid data points in the last 4 years ({valid_rev_points} found).")
    else:
         projection_log.append(f"  - Revenue Growth (3yr): Insufficient historical Revenue data ({len(historical_revenues_list)} years found, need 4).")

    # 2. Fallback to Default if Revenue Growth Failed
    if not found_growth:
        initial_growth_rate = default_init_growth # Use the 6.5% default
        growth_source = f"Default ({format_percent(default_init_growth)})"
        projection_log.append(f"⚠️ Revenue growth calculation failed. Using {growth_source}.")

    # 3. Apply Floor (1.1%) and Ceiling (15%)
    unbounded_rate = initial_growth_rate # Store original for comparison
    initial_growth_rate = max(initial_growth_rate, MIN_INITIAL_GROWTH_FLOOR) # Apply 1.1% floor
    initial_growth_rate = min(initial_growth_rate, MAX_INITIAL_GROWTH_CEILING) # Apply 15% ceiling

    if abs(initial_growth_rate - unbounded_rate) > 1e-6: # Check if bounding changed the value
        growth_source += " (Bounded)"
        projection_log.append(f"  Applied Bounds: Final Initial Growth Rate = {format_percent(initial_growth_rate)} (Floor: {format_percent(MIN_INITIAL_GROWTH_FLOOR)}, Ceiling: {format_percent(MAX_INITIAL_GROWTH_CEILING)})")
    else:
         projection_log.append(f"  Final Initial Growth Rate = {format_percent(initial_growth_rate)}")

    projection_results['initial_growth_rate'] = initial_growth_rate
    projection_results['growth_source'] = growth_source

    # --- Terminal Growth Rate Calculation (Remains the same conservative approach) ---
    # Add check for valid initial_growth_rate before using it as a cap
    # Use the final, bounded initial_growth_rate here
    g_cap_initial_valid = initial_growth_rate if pd.notna(initial_growth_rate) and math.isfinite(initial_growth_rate) else TERMINAL_GROWTH_RATE_CAP # Fallback if initial is invalid

    g_cap_config = TERMINAL_GROWTH_RATE_CAP; g_cap_wacc = disc_rate * 0.67; g_cap_rf = final_user_rf * 0.95 # Use final RF
    projection_log.append(f"--- Determining Terminal Growth Rate ---")
    projection_log.append(f"Terminal Growth Caps Considered: Config={format_percent(g_cap_config)}, WACC*0.67={format_percent(g_cap_wacc)}, Initial(Final)={format_percent(g_cap_initial_valid)}, Rf*0.95={format_percent(g_cap_rf)}")
    terminal_growth_rate = min(g_cap_config, g_cap_wacc, g_cap_initial_valid, g_cap_rf); projection_log.append(f"↳ Minimum of Caps: {format_percent(terminal_growth_rate)}")
    original_g_before_floor = terminal_growth_rate
    terminal_growth_rate = max(MIN_GROWTH, terminal_growth_rate) # Apply floor (MIN_GROWTH=0.00 usually)
    if abs(terminal_growth_rate - original_g_before_floor) > 1e-6 : projection_log.append(f"↳ Adjusted by Floor ({format_percent(MIN_GROWTH)}): {format_percent(terminal_growth_rate)}")
    if terminal_growth_rate >= disc_rate: original_g = terminal_growth_rate; terminal_growth_rate = disc_rate - 0.005; projection_log.append(f"⚠️ Terminal Growth ({format_percent(original_g)}) >= WACC ({format_percent(disc_rate)}). Reduced to {format_percent(terminal_growth_rate)}.")
    projection_log.append(f"▶️ Using Final Terminal Growth Rate (g): {format_percent(terminal_growth_rate)}"); projection_results['terminal_growth_rate'] = terminal_growth_rate

    # --- Project FCFFs using Linear Decay from initial_growth_rate -> terminal_growth_rate ---
    projected_fcffs = []; growth_rates_applied = []; pv_fcffs = []; cumulative_pv_fcff = 0.0
    current_fcff_for_proj = fcff_current # Start projection base from T-0 FCFF

    # Ensure effective initial growth is valid, fallback if needed (should be valid after default/bounding)
    if pd.isna(initial_growth_rate) or not math.isfinite(initial_growth_rate):
        projection_log.append(f"⚠️ FATAL: Invalid final initial growth rate ({initial_growth_rate}) despite checks. Using 0% for projection start.")
        effective_initial_growth = 0.0
    else:
        effective_initial_growth = initial_growth_rate # Use the final bounded rate

    effective_terminal_growth = terminal_growth_rate

    # Handle negative starting FCFF carefully
    if current_fcff_for_proj <= 0:
        projection_log.append("⚠️ Starting FCFF (T-0) <= 0. Applying minimal growth.")
        effective_initial_growth = min(effective_initial_growth, 0.01); # Minimal positive initial growth
        effective_terminal_growth = min(terminal_growth_rate, 0.00) # Minimal non-negative terminal

    projection_log.append("--- FCFF Projections ---"); projection_log.append(f"{'Year':<5} {'Growth':<8} {'FCFF':<18} {'PV Factor':<10} {'PV FCFF':<18}"); projection_log.append("-" * 60) # Adjusted width
    last_valid_fcff = current_fcff_for_proj # Use the starting FCFF (T-0) for year 1 growth calculation

    for year in range(1, forecast_yrs + 1):
        # Linear decay of growth from initial to terminal
        # Prevent division by zero if forecast_yrs is 0 or 1
        if forecast_yrs > 0:
             decay_factor = max(0, (forecast_yrs - year)) / forecast_yrs # Starts closer to initial, ends at terminal
        else:
             decay_factor = 0 # Goes straight to terminal if forecast_yrs = 0

        growth_rate = effective_terminal_growth + (effective_initial_growth - effective_terminal_growth) * decay_factor
        growth_rate = max(MIN_GROWTH, growth_rate) # Apply general floor (0%) to yearly growth too

        # Moderate growth/shrinkage if starting FCFF was negative
        if current_fcff_for_proj <= 0: growth_rate = max(-0.10, min(growth_rate, 0.05)) # Limit growth/shrinkage rate

        growth_rates_applied.append(growth_rate)
        fcff_forecast = last_valid_fcff * (1 + growth_rate) # Calculate next FCFF

        if not np.isfinite(fcff_forecast): fcff_forecast = 0 # Handle potential NaN
        projected_fcffs.append(fcff_forecast)
        last_valid_fcff = fcff_forecast # IMPORTANT: Update last_valid_fcff for the *next* iteration

        pv_factor = 1 / ((1 + disc_rate) ** year); pv_fcff = fcff_forecast * pv_factor
        if not np.isfinite(pv_fcff): pv_fcff = 0 # Handle potential NaN PV
        pv_fcffs.append(pv_fcff); cumulative_pv_fcff += pv_fcff # Add PV (even if 0)
        projection_log.append(f"{year:<5} {format_percent(growth_rate):<8} {format_currency(fcff_forecast, '', 0, is_indian=is_indian):<18} {pv_factor:<10.4f} {format_currency(pv_fcff, '', 0, is_indian=is_indian):<18}")

    projection_results['projected_fcffs'] = projected_fcffs; projection_results['growth_rates'] = growth_rates_applied; projection_results['pv_fcffs'] = pv_fcffs
    projection_results['cumulative_pv_fcff'] = cumulative_pv_fcff if np.isfinite(cumulative_pv_fcff) else 0.0 # Handle NaN sum
    projection_log.append("-" * 60); projection_log.append(f"Sum of PV(Forecasted FCFFs): {format_currency(cumulative_pv_fcff, symbol=currency_sym, is_indian=is_indian)}") # Use proper symbol here

    # Calculate Terminal Value (Remains the same)
    terminal_value = np.nan; pv_terminal_value = np.nan
    final_forecast_fcff = projected_fcffs[-1] if projected_fcffs and np.isfinite(projected_fcffs[-1]) else 0

    if (disc_rate - terminal_growth_rate) <= 1e-6 or not np.isfinite(terminal_growth_rate):
         projection_log.append(f"❌ ERROR: WACC ({format_percent(disc_rate)}) <= Terminal Growth ({format_percent(terminal_growth_rate)}) or g invalid. TV calc failed."); projection_results['error'] = "WACC <= g or g invalid. TV Calc failed."
    elif not np.isfinite(final_forecast_fcff):
         projection_log.append(f"❌ ERROR: Final projected FCFF (Year {forecast_yrs}) is invalid. TV calc failed."); projection_results['error'] = "Final FCFF invalid. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    else:
         if final_forecast_fcff <= 0:
             projection_log.append(f"ℹ️ FCFF Year {forecast_yrs} ({format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)}) <= 0. Setting TV=0."); # Pass flag
             terminal_value = 0.0; pv_terminal_value = 0.0
         else:
              fcff_n_plus_1 = final_forecast_fcff * (1 + terminal_growth_rate)
              terminal_value = fcff_n_plus_1 / (disc_rate - terminal_growth_rate)
              if not np.isfinite(terminal_value) or terminal_value < 0:
                  projection_log.append(f"⚠️ Calculated TV invalid/negative ({format_currency(terminal_value, currency_sym, is_indian=is_indian)}). Setting TV=0.") # Check validity post-calc
                  terminal_value = 0.0; pv_terminal_value = 0.0
              else:
                   pv_factor_tv = 1 / ((1 + disc_rate) ** forecast_yrs); pv_terminal_value = terminal_value * pv_factor_tv
                   projection_log.append(f"FCFF Yr {forecast_yrs}: {format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)} -> FCFF Yr {forecast_yrs+1}: {format_currency(fcff_n_plus_1, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"Terminal Value (Yr {forecast_yrs}): {format_currency(terminal_value, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"PV Factor TV: {pv_factor_tv:.4f} -> PV(TV): {format_currency(pv_terminal_value, currency_sym, is_indian=is_indian)}")

    projection_results['terminal_value'] = terminal_value if np.isfinite(terminal_value) else 0.0
    projection_results['pv_terminal_value'] = pv_terminal_value if np.isfinite(pv_terminal_value) else 0.0 # Handle final NaN

    display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
    return projection_results

# Execute FCFF calculation and projection
if np.isfinite(discount_rate):
    # ----> Pass currency symbol, flag, NEW default growth rate. No long history needed now. <----
    fcff_results = calculate_and_project_fcff(
        stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'], # Removed history_long
        wacc_components, discount_rate, FORECAST_YEARS, risk_free_rate_input, DEFAULT_GROWTH_RATE,
        currency_symbol, is_indian_stock)
else:
    display(Markdown("❌ **Skipping FCFF Projection due to invalid WACC.**")); fcff_results = {'error': 'Invalid WACC.'}


# %% [markdown]
# ## 7. Enterprise Value and Equity Value Calculation

# %%
# Valuation Function Definition (No changes here)
def calculate_final_valuation(fcff_data, balance_sheet_df, shares_outstanding_val, currency_sym, wacc_data, is_indian): # Added flag
    display(Markdown("### Final Valuation Calculation..."))
    valuation = {'enterprise_value': np.nan, 'equity_value': np.nan, 'intrinsic_value_per_share': np.nan, 'error': None}
    valuation_log = []
    if fcff_data is None or fcff_data.get('error'):
        valuation['error'] = f"FCFF Error: {fcff_data.get('error', 'Unknown')}"; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    pv_forecast_sum = fcff_data.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_data.get('pv_terminal_value', np.nan)
    if not np.isfinite(pv_forecast_sum) or not np.isfinite(pv_tv):
         valuation['error'] = f"Invalid PV components (PV FCFF: {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}, PV TV: {format_currency(pv_tv, currency_sym, is_indian=is_indian)})."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation # Pass flag

    enterprise_value = pv_forecast_sum + pv_tv; valuation['enterprise_value'] = enterprise_value
    # ----> Pass is_indian flag to format_currency calls <----
    valuation_log.append(f"PV(Forecast FCFFs): {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"PV(Terminal Value):   {format_currency(pv_tv, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"↳ Enterprise Value (EV): {format_currency(enterprise_value, currency_sym, is_indian=is_indian)}")

    # Get Debt & Cash from BS (use Debt from WACC for consistency)
    total_debt_latest = 0.0; cash_and_equivalents = 0.0; minority_interest = 0.0
    debt_source_info = wacc_data.get('debt_calc_method', '?'); total_debt_latest = wacc_data.get('debt_value_used', 0.0) # Get debt used in WACC

    # --- CORRECTED INDENTATION WITHIN IF/ELSE (Remains Correct) ---
    if not balance_sheet_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0] # Standard indent (8 spaces)
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}") # Standard indent
        cash_keys = ["Cash And Cash Equivalents", "Cash Cash Equivalents And Short Term Investments", "Cash Financial", "Cash", "Cash And Short Term Investments"] # Standard indent
        cash_and_equivalents = safe_get(latest_bs, cash_keys, "BS", default=0) # Standard indent
        valuation_log.append(f"  + Cash & Equiv (BS): {format_currency(cash_and_equivalents, currency_sym, is_indian=is_indian)}") # Standard indent
        minority_interest = safe_get(latest_bs, ["Minority Interest", "Noncontrolling Interest"], "BS", default=0) # Standard indent
        if minority_interest != 0.0: # Nested if, standard indent (12 spaces)
            valuation_log.append(f"  - Minority Interest: {format_currency(minority_interest, currency_sym, is_indian=is_indian)}")
    else:
        valuation_log.append("⚠️ Balance Sheet empty.") # Standard indent (8 spaces)
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}") # Standard indent
        valuation_log.append(f"  + Cash & Equiv: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)") # Standard indent
    # --- END CORRECTION ---

    if not all(np.isfinite(v) for v in [enterprise_value, total_debt_latest, cash_and_equivalents, minority_interest]):
         valuation['error'] = "Invalid component for EV->Equity bridge."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    equity_value = enterprise_value - total_debt_latest + cash_and_equivalents - minority_interest; valuation['equity_value'] = equity_value
    # ----> Pass is_indian flag to format_currency calls <----
    valuation_log.append(f"↳ Equity Value: {format_currency(equity_value, currency_sym, is_indian=is_indian)}")

    # Calculate Intrinsic Value Per Share
    intrinsic_value_per_share = np.nan
    if not np.isfinite(equity_value): valuation['error'] = f"Equity value invalid ({equity_value})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    elif shares_outstanding_val <= 0: valuation['error'] = f"Shares outstanding invalid ({shares_outstanding_val})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    else:
        intrinsic_value_per_share = equity_value / shares_outstanding_val;
        # Format shares with commas, independent of currency
        valuation_log.append(f"  ÷ Shares Out: {shares_outstanding_val:,.0f}")
        # ----> Pass is_indian flag to format_currency calls <----
        valuation_log.append(f"▶️ Intrinsic Value/Share: {format_currency(intrinsic_value_per_share, currency_sym, is_indian=is_indian)}")
    valuation['intrinsic_value_per_share'] = intrinsic_value_per_share if np.isfinite(intrinsic_value_per_share) else np.nan

    display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
    return valuation

# Execute Final Valuation
if 'error' not in fcff_results or fcff_results['error'] is None:
    # ----> Pass flag <----
    final_valuation = calculate_final_valuation(fcff_results, stock_data['balance_sheet'], shares_outstanding, currency_symbol, wacc_components, is_indian_stock)
else:
    display(Markdown("❌ **Skipping Final Valuation due to FCFF errors.**")); final_valuation = {'error': fcff_results.get('error', 'Upstream FCFF error.')}


# %% [markdown]
# ## 8. Financial Ratio Analysis

# %%
# Ratio Function Definition (No changes here)
def analyze_ratios(stock_info_dict, financials_df, balance_sheet_df, cashflow_df,
                   enterprise_val, current_price_val, shares_outstanding_val,
                   currency_sym, user_benchmark_dict, # Only user inputs
                   benchmark_source_dict, # Dict tracking source per ratio
                   wacc_data, is_indian): # Pass WACC data for consistent debt and flag

    """ Calculates ratios, compares to user/fallback benchmarks, uses WACC-consistent debt. """
    display(Markdown("### Financial Ratio Analysis..."))
    ratios = {} # Calculated ratios {RatioName: value}
    ratio_log = []
    ratio_data_for_table = [] # For HTML table display
    ratio_plotly_data = {'Ratio': [], 'Company': [], 'Benchmark': [], 'Benchmark Source': [], 'Status': []} # For Plotly chart
    company_nm = stock_info_dict.get("shortName", ticker_symbol)

    # --- Pre-checks ---
    valid_price = current_price_val is not None and np.isfinite(current_price_val) and current_price_val > 0
    valid_shares = shares_outstanding_val is not None and np.isfinite(shares_outstanding_val) and shares_outstanding_val > 0
    valid_enterprise_val = enterprise_val is not None and np.isfinite(enterprise_val)

    # Check for essential statements
    if financials_df.empty or balance_sheet_df.empty:
         ratio_log.append("❌ ERROR: Income Statement or Balance Sheet empty. Cannot calculate most ratios.")
         display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
         return {}, [] # Return empty results

    latest_fin = financials_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0] if not cashflow_df.empty else pd.Series(dtype=float)

    # --- Calculate Ratio Components (using safe_get with NaN default where appropriate) ---
    net_income = safe_get(latest_fin, ["Net Income Continuous Operations", "Net Income", "Net Income Common Stockholders", "Net Income Applicable To Common Shares"], default=np.nan)
    revenue = safe_get(latest_fin, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
    depreciation = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0) # Default 0 ok here
    ebitda = (ebit + depreciation) if np.isfinite(ebit) and np.isfinite(depreciation) else np.nan

    total_equity = safe_get(latest_bs, ["Total Stockholder Equity", "Stockholders Equity", "Total Equity Gross Minority Interest", "Common Stock Equity"], default=np.nan)
    total_assets = safe_get(latest_bs, ["Total Assets"], default=np.nan)
    current_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=np.nan)
    current_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=np.nan)
    inventory = safe_get(latest_bs, ["Inventory"], default=0) # Default 0 ok here

    # --- Use Debt value consistent with WACC calculation ---
    total_debt_ratio = wacc_data.get('debt_value_used', 0.0); debt_source_info = wacc_data.get('debt_calc_method', '?');
    # ----> Pass is_indian flag to format_currency calls <----
    ratio_log.append(f"Using Debt = {format_currency(total_debt_ratio, currency_sym, is_indian=is_indian)} (Source: WACC - {debt_source_info})")

    # Per Share Metrics
    eps = (net_income / shares_outstanding_val) if valid_shares and np.isfinite(net_income) and net_income != 0 else np.nan
    bvps = (total_equity / shares_outstanding_val) if valid_shares and np.isfinite(total_equity) and total_equity != 0 else np.nan

    # Capital Employed (ROCE denominator)
    capital_employed = np.nan; ce_source = "N/A"
    # Definition 1: Total Assets - Current Liabilities
    if np.isfinite(total_assets) and np.isfinite(current_liab):
        capital_employed_def1 = total_assets - current_liab
        # Check if calculated value is valid (positive)
        if np.isfinite(capital_employed_def1) and capital_employed_def1 > 1.0: # Use 1.0 as a small threshold
            capital_employed = capital_employed_def1
            ce_source = "A-CL"

    # Definition 2: Total Equity + Total Debt (Fallback)
    if not np.isfinite(capital_employed): # If Def1 failed or was invalid, try Def2
        if np.isfinite(total_equity) and np.isfinite(total_debt_ratio):
             capital_employed_def2 = total_equity + total_debt_ratio
             if np.isfinite(capital_employed_def2) and capital_employed_def2 > 1.0: # Use 1.0 threshold
                 capital_employed = capital_employed_def2
                 ce_source = "E+D" # Update source if fallback used

    # Check if still invalid after fallbacks
    if not np.isfinite(capital_employed):
         ce_source = "Invalid/Unavailable"
         ratio_log.append("⚠️ Capital Employed could not be reliably calculated. ROCE affected.")

    # --- Calculate Ratios (Result is NaN if components invalid/zero denominator) ---
    ratios['P/E'] = (current_price_val / eps) if valid_price and np.isfinite(eps) and eps != 0 else np.nan
    ratios['P/B'] = (current_price_val / bvps) if valid_price and np.isfinite(bvps) and bvps != 0 else np.nan
    ratios['EV/EBITDA'] = (enterprise_val / ebitda) if valid_enterprise_val and np.isfinite(ebitda) and ebitda != 0 else np.nan
    ratios['ROE'] = (net_income / total_equity) if np.isfinite(net_income) and np.isfinite(total_equity) and total_equity != 0 else np.nan
    ratios['ROA'] = (net_income / total_assets) if np.isfinite(net_income) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['ROCE'] = (ebit / capital_employed) if np.isfinite(ebit) and np.isfinite(capital_employed) and capital_employed != 0 else np.nan # Uses calculated CE
    ratios['Asset Turnover'] = (revenue / total_assets) if np.isfinite(revenue) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['Current Ratio'] = (current_assets / current_liab) if np.isfinite(current_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan
    quick_assets = (current_assets - inventory) if np.isfinite(current_assets) and np.isfinite(inventory) else np.nan
    ratios['Quick Ratio'] = (quick_assets / current_liab) if np.isfinite(quick_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan

    # --- Compare with User/Fallback Benchmarks & Prepare Data Tables ---
    display(Markdown("*(Comparing ratios to user/fallback benchmarks...)*"))
    for key in REQUIRED_BENCHMARK_RATIOS:
        company_value = ratios.get(key); is_percent_type = key in ['ROE', 'ROA', 'ROCE']
        # Handle formatting, including potential Inf values
        # NOTE: Ratios themselves are typically NOT formatted as currency (Lakh/Crore)
        # So we use format_currency with '' symbol and is_indian=False intentionally here
        value_fmt = format_percent(company_value, 1) if is_percent_type else format_currency(company_value, '', 2, is_indian=False)
        if value_fmt == "N/A" and company_value is not None and not math.isfinite(company_value): value_fmt = "Inf" # Explicitly show Inf

        # Determine which benchmark to use and its source
        user_benchmark = user_benchmark_dict.get(key, None); fallback_benchmark = DEFAULT_BENCHMARK_RATIOS.get(key, None)
        benchmark_source = benchmark_source_dict.get(key, "?"); final_benchmark = None
        if benchmark_source == 'User Input' and user_benchmark is not None: final_benchmark = user_benchmark
        elif benchmark_source == 'Fallback Default' and fallback_benchmark is not None: final_benchmark = fallback_benchmark
        else: benchmark_source = "Not Available" # If neither user nor default was set

        # Format benchmark for display
        bench_fmt = "N/A"
        if final_benchmark is not None: bench_fmt = f"{final_benchmark:.1f}%" if is_percent_type else f"{final_benchmark:.2f}"; bench_fmt += f" <i style='font-size:0.8em; color:grey'>({benchmark_source.replace('Fallback ','')})</i>"
        else: bench_fmt = "<i style='color:grey'>Not Provided</i>"

        # Determine status vs benchmark
        status = "<i style='color:grey'>N/A</i>"; status_color = "grey"
        if np.isfinite(company_value) and final_benchmark is not None and np.isfinite(final_benchmark):
            numeric_value = company_value; numeric_benchmark = final_benchmark
            good_color = '#2ecc71'; bad_color = '#e74c3c'; neutral_color = '#f39c12'; grey_color='#7f8c8d'
            # Handle potential division by zero for ratio diff
            if abs(numeric_benchmark) > 1e-9: ratio_diff = (numeric_value - numeric_benchmark) / abs(numeric_benchmark)
            else: ratio_diff = float('inf') if numeric_value > 0 else float('-inf') if numeric_value < 0 else 0.0
            # Set status based on ratio type and difference threshold (e.g., +/- 20%)
            if key in ['ROE', 'ROA', 'ROCE', 'Asset Turnover', 'Current Ratio', 'Quick Ratio']: # Higher is better
                 if ratio_diff > 0.20: status_color = good_color; status = f"⬆️ Strong"
                 elif ratio_diff < -0.20: status_color = bad_color; status = f"⬇️ Weak"
                 else: status_color = grey_color; status = f"~ Average"
            elif key in ['P/E', 'P/B', 'EV/EBITDA']: # Lower might be better (valuation) or neutral
                 if ratio_diff < -0.20: status_color = good_color; status = f"✅ Low"
                 elif ratio_diff > 0.20: status_color = neutral_color; status = f"⚠️ High" # High valuation isn't always bad, but notable
                 else: status_color = grey_color; status = f"~ Average"
            status = f"<span style='color:{status_color}'>{status}</span>"
            # Prepare data for Plotly chart (convert % ratios back to % for plotting if needed)
            plot_value = numeric_value * 100 if is_percent_type else numeric_value; plot_benchmark = numeric_benchmark # Benchmark already in correct format
            ratio_plotly_data['Ratio'].append(key); ratio_plotly_data['Company'].append(plot_value); ratio_plotly_data['Benchmark'].append(plot_benchmark);
            ratio_plotly_data['Benchmark Source'].append(benchmark_source); ratio_plotly_data['Status'].append(status) # Store HTML status for potential hover
        elif not np.isfinite(company_value): status = "<i style='color:grey'>Company N/A</i>"
        elif final_benchmark is None: status = "<i style='color:grey'>No Benchmark</i>"
        ratio_data_for_table.append([f"<b>{key}</b>", value_fmt, bench_fmt, status])

    # --- Display Logs & Table ---
    if ratio_log: display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
    if ratio_data_for_table:
        header = ["Ratio", f"{company_nm} Value", "Benchmark (Source)", "Status"]; html_table = "<p>(*Status vs User/Default benchmarks. Threshold: +/- 20%.)</p>"; html_table += "<table border='1' style='border-collapse:collapse; width:98%; margin-left:auto; margin-right:auto; font-size: 0.9em;'>"; html_table += "<thead style='background-color:#e9ecef; color: #495057;'><tr>" + "".join(f"<th style='padding: 8px; text-align:center;'>{h}</th>" for h in header) + "</tr></thead>"; html_table += "<tbody>"
        for i, row in enumerate(ratio_data_for_table): bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff"; html_table += f"<tr style='background-color:{bg_color};'>"; html_table += f"<td style='padding: 6px; text-align:left;'>{row[0]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[1]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[2]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[3]}</td>"; html_table += "</tr>"
        html_table += "</tbody></table><br>"; display(HTML(html_table))
    else: display(Markdown("*No ratio data available.*"))

    # --- Display Ratio Comparison Chart (Plotly Bullet) ---
    if ratio_plotly_data['Ratio']:
       num_ratios = len(ratio_plotly_data['Ratio']); fig_ratios = go.Figure()
       def get_range_colors(k): return ["#fadbd8", "#fdebd0", "#d5f5e3"] if k in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio'] else ["#d5f5e3","#fdebd0","#fadbd8"] if k in ['P/E','P/B','EV/EBITDA'] else ['lightgrey','darkgrey','grey'] # Bad/Ok/Good vs Good/Ok/Bad
       for i in range(num_ratios):
           ratio_name = ratio_plotly_data['Ratio'][i]; company_val = ratio_plotly_data['Company'][i]; benchmark_val = ratio_plotly_data['Benchmark'][i]; bench_src_plot = ratio_plotly_data['Benchmark Source'][i]
           # Skip plotting if data invalid
           if not np.isfinite(company_val) or not np.isfinite(benchmark_val): print(f"Skipping plot for {ratio_name}: Invalid Value ({company_val}) or Bench ({benchmark_val})."); continue

           range_colors = get_range_colors(ratio_name); is_percent = ratio_name in ['ROE', 'ROA', 'ROCE']
           # Define poor/good factors relative to benchmark (handle zero/negative benchmarks)
           poor_f=0.8; good_f=1.2;
           range1_limit = benchmark_val * (poor_f if benchmark_val >=0 else good_f) # Lower band for +ve, upper for -ve
           range2_limit = benchmark_val * (good_f if benchmark_val >=0 else poor_f) # Upper band for +ve, lower for -ve
           # Dynamic axis range setting
           abs_max = max(abs(company_val), abs(benchmark_val)); plot_max_val = abs_max * 1.7 + 1; plot_min_val = 0
           if company_val < 0 or benchmark_val < 0: plot_min_val = min(company_val, benchmark_val) * 1.7 - 1 # Extend lower if negative values present
           # Apply some sensible caps to axis ranges based on ratio type
           if ratio_name in ['P/E', 'EV/EBITDA']: plot_max_val = min(plot_max_val, 100); plot_min_val = max(plot_min_val, -20)
           elif ratio_name in ['P/B']: plot_max_val = min(plot_max_val, 20); plot_min_val = max(plot_min_val, -5)
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: plot_max_val = min(plot_max_val, 5); plot_min_val = max(plot_min_val, 0)
           elif is_percent: plot_max_val = min(plot_max_val, 75); plot_min_val= max(plot_min_val,-50)
           else: plot_max_val = min(plot_max_val, 1000) # Generic cap
           if plot_min_val >= plot_max_val: plot_max_val = plot_min_val + 1 # Ensure max > min

           # Define gauge steps based on calculated ranges
           steps = [];
           if abs(benchmark_val) < 1e-9: steps = [{'range': [plot_min_val, plot_max_val], 'color':'lightgrey'}] # Handle zero benchmark case
           else: step_ranges = sorted([plot_min_val, range1_limit, range2_limit, plot_max_val]); step_colors = range_colors if benchmark_val >= 0 else range_colors[::-1]; steps = [{'range': [step_ranges[j], step_ranges[j+1]], 'color': step_colors[j]} for j in range(3)]

           # Format title and delta colors based on whether lower is better
           bench_fmt_title = f"{benchmark_val:.1f}%" if is_percent else f"{benchmark_val:.2f}"; title_text = f"<b>{ratio_name}</b><br><i style='font-size:0.8em'>(Bench: {bench_fmt_title} [{bench_src_plot.replace('Fallback ','')}])</i>"
           lower_is_better = ratio_name in ['P/E', 'P/B', 'EV/EBITDA']; delta_decreasing_color = good_color if lower_is_better else bad_color; delta_increasing_color = bad_color if lower_is_better else good_color
           fig_ratios.add_trace(go.Indicator(
                mode = "number+gauge+delta", value = company_val, number={'suffix': "%" if is_percent else "", 'valueformat': ".1f"}, # Format number display
                delta = {'reference': benchmark_val, 'relative': True, 'valueformat':".0%", 'decreasing': {'color': delta_decreasing_color}, 'increasing': {'color': delta_increasing_color}},
                domain = {'row': i, 'column': 0}, title = {'text' : title_text, 'font':{'size':11}},
                gauge = { 'shape': "bullet", 'axis': {'range': [plot_min_val, plot_max_val] if plot_min_val < plot_max_val else [plot_min_val, plot_min_val+1], 'tickformat':'.1f'}, # Use calculated axis range
                    'threshold': { 'line': {'color': "black", 'width': 2}, 'thickness': 0.8, 'value': benchmark_val }, # Show benchmark line
                    'steps': [ s for s in steps if s and isinstance(s.get('range'), list) and len(s['range'])==2 and s['range'][0] < s['range'][1] ], # Use calculated steps, ensure valid
                    'bar': {'color': "#2c3e50", 'thickness': 0.5} }
            ))
       fig_ratios.update_layout(height = max(300, num_ratios * 85), grid = {'rows': num_ratios, 'columns': 1, 'pattern': "independent"}, title_text='Ratio Performance vs. Benchmarks', title_x=0.5, margin=dict(l=160, r=40, t=60, b=30))
       fig_ratios.show()
    else: display(Markdown("*No valid ratios/benchmarks for comparison chart.*"))

    return ratios, ratio_data_for_table

# Execute Ratio Analysis
if 'error' not in final_valuation or final_valuation['error'] is None:
    # ----> Pass flag <----
    ratios_calculated, ratio_table = analyze_ratios(
        stock_data['info'], stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'],
        final_valuation.get('enterprise_value'), current_price, shares_outstanding, currency_symbol,
        user_benchmarks, benchmark_sources, wacc_components, is_indian_stock )
else: display(Markdown("❌ **Skipping Ratio Analysis due to prior errors.**")); ratios_calculated, ratio_table = {}, []

# %% [markdown]
# ## 9. Summary & Visualization Dashboard

# %%
# Dashboard Function Definition (No changes here)
def display_summary_dashboard(company_nm, ticker_sym, current_pr, intrinsic_val_per_sh,
                              wacc_val, term_growth, init_growth, growth_src,
                              final_val_dict, hist_df, fcff_res, currency_sym, info_dict, input_src_dict,
                              is_indian): # Added flag

    display(Markdown(f"## Valuation Summary Dashboard: {company_nm} ({ticker_sym})"))
    source_summary = " | ".join(f"{k}:`{v.split('(')[0].strip()}`" for k, v in input_src_dict.items())
    # Display the source of the initial growth rate
    display(Markdown(f"**Input Sources:** {source_summary} | Growth:`{growth_src}`")) # Show full source string

    # Check validity of key inputs for display logic
    valid_intrinsic = intrinsic_val_per_sh is not None and np.isfinite(intrinsic_val_per_sh) # Allow <= 0 intrinsic
    valid_current_pr = current_pr is not None and np.isfinite(current_pr) and current_pr > 0
    valid_wacc = wacc_val is not None and np.isfinite(wacc_val)
    valid_term_g = term_growth is not None and np.isfinite(term_growth)
    valid_init_g = init_growth is not None and np.isfinite(init_growth)

    # --- Indicator Subplots ---
    fig_indicators = make_subplots(rows=1, cols=3, specs=[[{'type': 'indicator'}, {'type': 'indicator'}, {'type': 'indicator'}]],
                                 subplot_titles=("Price vs. Intrinsic", "WACC", "Growth Rates"), horizontal_spacing=0.1, column_widths=[0.4, 0.3, 0.3])

    # Indicator 1: Price vs Intrinsic
    intrinsic_val_num = intrinsic_val_per_sh if valid_intrinsic else np.nan
    # ----> Pass flag <----
    intrinsic_val_fmt = format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian) if valid_intrinsic else "N/A"
    current_pr_num = current_pr if valid_current_pr else 0.0; upside_fmt = ""; delta_config = None
    if valid_intrinsic and valid_current_pr :
        if intrinsic_val_num > 0: upside_pct = (intrinsic_val_num / current_pr_num) - 1; upside_fmt = f" ({upside_pct:+.1%})"
        comparison_text = f"Intrinsic: {intrinsic_val_fmt}{upside_fmt}"
        # Delta prefix needs to adapt - use symbol directly
        delta_config = {'reference': intrinsic_val_num, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym, 'decreasing': {'color': '#2ecc71', 'symbol': '▲ Upside'}, 'increasing': {'color': '#e74c3c', 'symbol': '▼ Downside'}} # Delta vs Intrinsic
    elif valid_intrinsic: comparison_text = f"Intrinsic: {intrinsic_val_fmt} (Current N/A)"
    else: comparison_text = f"Intrinsic: N/A"
    # Number prefix needs to adapt
    fig_indicators.add_trace(go.Indicator(mode = "number+delta" if delta_config else "number", value = current_pr_num if valid_current_pr else 0, number={'prefix': currency_sym + " " if is_indian else currency_sym, 'valueformat': '.2f', 'font':{'size':24}}, title = {"text": f"<span style='font-size:0.8em'>{comparison_text}</span>"}, delta = delta_config, domain = {'row': 0, 'column': 0}), row=1, col=1)

    # Indicator 2: WACC Gauge
    wacc_val_num = wacc_val * 100 if valid_wacc else None # Use None if invalid
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = wacc_val_num, # Pass None value if invalid
        number = {'suffix': "%",
                  'valueformat':'.1f' if valid_wacc else '', # Only format if valid
                  'font':{'size':20}}, # REMOVED INVALID 'text' property here
        gauge = {'axis': {'range': [MIN_WACC*100, MAX_WACC*100]}, 'bar': {'color': "darkblue" if valid_wacc else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc", 'steps' : [ {'range': [MIN_WACC*100, 9], 'color': "#d5f5e3"}, {'range': [9, 14], 'color': "#fdebd0"}, {'range': [14, MAX_WACC*100], 'color': "#fadbd8"}]} if valid_wacc else None, # Conditional gauge
        title = {"text": "<span style='font-size:0.8em'>WACC</span>"}, domain = {'row': 0, 'column': 1}), row=1, col=2)

    # Indicator 3: Growth Gauge
    term_growth_num = term_growth * 100 if valid_term_g else None; init_growth_fmt = format_percent(init_growth, 1) if valid_init_g else "N/A"
    growth_title_text = f"<span style='font-size:0.8em'>Terminal Growth (g)<br>(Initial: {init_growth_fmt})</span>"
    # Dynamic range for growth gauge
    gauge_max_range = max(TERMINAL_GROWTH_RATE_CAP * 100 + 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) + 1)
    gauge_min_range = MIN_GROWTH * 100 - 1
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = term_growth_num, # Pass None value if invalid
         number = {'suffix': "%",
                   'valueformat':'.1f' if valid_term_g else '', # Only format if valid
                   'font':{'size':20}},
         gauge = {'axis': {'range': [gauge_min_range, gauge_max_range], 'tickwidth': 1}, 'bar': {'color': "darkgrey" if valid_term_g else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [{'range': [gauge_min_range, 2.0], 'color': "lightcoral"}, {'range': [2.0, 4.5], 'color': "lightyellow"}, {'range': [4.5, gauge_max_range], 'color': "lightgreen"}],
                 'threshold' : {'line': {'color': "black", 'width': 3}, 'thickness': 0.75, 'value': term_growth_num } if valid_term_g else None } if valid_term_g else None, # Conditional gauge/threshold
         title = {"text": growth_title_text}, domain = {'row': 0, 'column': 2}), row=1, col=3)

    fig_indicators.update_layout(height=220, margin=dict(l=30, r=30, t=50, b=20)); fig_indicators.show()

    # --- Historical Price Chart ---
    if not hist_df.empty:
        # ----> Use currency symbol in axis title <----
        fig_hist = px.line(hist_df, x=hist_df.index, y='Close', title=f'{company_nm} Price History vs. DCF'); fig_hist.update_layout(xaxis_title='Date', yaxis_title=f'Price ({currency_sym})', height=400, legend_title_text='Legend'); fig_hist.update_xaxes(rangeslider_visible=False)
        # ----> Pass flag to format_currency in annotations <----
        if valid_intrinsic: fig_hist.add_hline(y=intrinsic_val_num, line_dash="dash", line_color="#e74c3c", opacity=0.8, annotation_text=f"DCF Intrinsic: {format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian)}", annotation_position="bottom right", annotation_font_color="#e74c3c")
        if valid_current_pr: fig_hist.add_hline(y=current_pr_num, line_dash="dot", line_color="#3498db", opacity=0.8, annotation_text=f"Current: {format_currency(current_pr_num, currency_sym, is_indian=is_indian)}", annotation_position="top left", annotation_font_color="#3498db")
        fig_hist.show()
    else: display(Markdown("*Historical price data unavailable.*"))

    # --- DCF EV Breakdown Pie Chart ---
    pv_forecast_sum = fcff_res.get('cumulative_pv_fcff', np.nan) if fcff_res else np.nan; pv_tv = fcff_res.get('pv_terminal_value', np.nan) if fcff_res else np.nan
    valid_pvs = np.isfinite([pv_forecast_sum, pv_tv]); total_ev = (pv_forecast_sum + pv_tv) if all(valid_pvs) else np.nan
    if np.isfinite(total_ev) and total_ev != 0 : # EV Breakdown Pie Chart
        labels_pie = []; values_pie = []; pull_pie = []; colors = ['#4e79a7','#f28e2b']
        # Only add components if they are non-zero (handle potential zero TV or PV FCFFs)
        if abs(pv_forecast_sum) > 1e-6: labels_pie.append(f'PV({FORECAST_YEARS}yr FCFFs)'); values_pie.append(pv_forecast_sum); pull_pie.append(0) # Check abs value
        if abs(pv_tv) > 1e-6: labels_pie.append('PV(TV)'); values_pie.append(pv_tv); pull_pie.append(0.05 if len(labels_pie) > 1 else 0) # Check abs value
        if labels_pie: # Only plot if there's something to plot
            # Format hover template based on currency
            hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>" # Default
            if is_indian:
                # Can't easily do dynamic Lakh/Crore in hovertemplate, use standard Rupee format
                 hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
            # ----> Pass flag <----
            fig_dcf_pie = go.Figure(data=[go.Pie(labels=labels_pie, values=values_pie, hole=.4, pull=pull_pie, marker_colors=colors[:len(labels_pie)], hovertemplate=hover_format, textinfo='percent+label', insidetextorientation='radial', sort=False)])
            fig_dcf_pie.update_layout( title_text=f"DCF EV Breakdown (Total EV: {format_currency(total_ev, currency_sym, is_indian=is_indian)})", title_x=0.5, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), height=350, margin=dict(t=60, b=50, l=20, r=20) )
            fig_dcf_pie.show()
        else: display(Markdown(f"*DCF components zero/negligible. Cannot generate EV Breakdown chart.*"))
    else: display(Markdown(f"*EV calculation failed ({format_currency(total_ev, currency_sym, is_indian=is_indian)}). Cannot generate EV Breakdown chart.*")) # Added value for debug

    # --- FCFF Projection Bar Chart ---
    if fcff_res and fcff_res.get('projected_fcffs'):
        years = [f'Y{i+1}' for i in range(len(fcff_res['projected_fcffs']))]; pv_list_clean = [p if np.isfinite(p) else 0 for p in fcff_res.get('pv_fcffs', [])]; fcff_list_clean = [f if np.isfinite(f) else 0 for f in fcff_res.get('projected_fcffs', [])]
        if len(pv_list_clean) == len(years) and len(fcff_list_clean) == len(years):
            df_fcff_plot = pd.DataFrame({'Year': years, 'Forecasted FCFF': fcff_list_clean, 'PV of FCFF': pv_list_clean}); df_fcff_plot['Forecasted FCFF'] = pd.to_numeric(df_fcff_plot['Forecasted FCFF']); df_fcff_plot['PV of FCFF'] = pd.to_numeric(df_fcff_plot['PV of FCFF'])
            df_fcff_melt = df_fcff_plot.melt(id_vars='Year', var_name='Cash Flow Type', value_name='Value')
            # ----> Use currency symbol in axis label <----
            fig_fcff_bar = px.bar(df_fcff_melt, x='Year', y='Value', color='Cash Flow Type', title='FCFF Forecast & Present Values', labels={'Value':f'Value ({currency_sym})'}, barmode='group', height=350, color_discrete_map={'Forecasted FCFF':'#5dade2', 'PV of FCFF':'#af7ac5'})
            # Format hover template based on currency
            hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>" # Default
            if is_indian:
                 hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>" # Cannot easily do Lakh/Crore here
            fig_fcff_bar.update_traces(hovertemplate=hover_bar_format); fig_fcff_bar.update_layout(margin=dict(t=60, b=30, l=50, r=30)); fig_fcff_bar.show()
        else: display(Markdown("*FCFF projection data mismatch/invalid. Cannot plot.*"))
    else: display(Markdown("*FCFF projection data unavailable.*"))

    # --- Valuation Conclusion ---
    display(Markdown("### Valuation Conclusion")); intrinsic_val = final_val_dict.get('intrinsic_value_per_share') if final_val_dict else np.nan
    current_pr_num = current_pr if valid_current_pr else np.nan; conclusion = ""
    if final_val_dict is None or final_val_dict.get('error'): conclusion = f"<span style='color:red'>❌ Valuation Error: {final_val_dict.get('error', '?')}</span>"
    # ----> Pass flag to format_currency calls <----
    elif not np.isfinite(intrinsic_val) or intrinsic_val <= 0: conclusion = f"⚠️ Valuation resulted in invalid/non-positive Intrinsic Value ({format_currency(intrinsic_val, currency_sym, is_indian=is_indian)})."
    elif not np.isfinite(current_pr_num): conclusion = f"✅ Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}. Cannot compare (Current Price N/A)."
    else:
         diff_pct = (intrinsic_val / current_pr_num) - 1
         current_fmt = format_currency(current_pr_num, currency_sym, is_indian=is_indian)
         intrinsic_fmt = format_currency(intrinsic_val, currency_sym, is_indian=is_indian)
         if abs(diff_pct) < 0.10: conclusion = f"<span style='color:grey'>~ Fairly Valued:</span> Price ({current_fmt}) vs DCF ({intrinsic_fmt}). Diff: {diff_pct:+.1%}"
         elif diff_pct < 0: overval_pct = (current_pr_num / intrinsic_val) - 1; conclusion = f"<span style='color:#e74c3c'>🔻 Potentially Overvalued:</span> Price ({current_fmt}) is **{overval_pct:+.1%}** above DCF ({intrinsic_fmt})."
         else: conclusion = f"<span style='color:#2ecc71'>✅ Potentially Undervalued:</span> Price ({current_fmt}) suggests **{diff_pct:+.1%}** upside to DCF ({intrinsic_fmt})."
    display(Markdown(f"> **{conclusion}**"))

# Execute Summary Dashboard Generation
if 'error' not in final_valuation or final_valuation['error'] is None:
    # ----> Pass flag <----
    display_summary_dashboard(
        company_name, ticker_symbol, current_price, final_valuation.get('intrinsic_value_per_share'),
        discount_rate, fcff_results.get('terminal_growth_rate'), fcff_results.get('initial_growth_rate'),
        fcff_results.get('growth_source', 'N/A'), final_valuation, stock_data['history'], fcff_results,
        currency_symbol, stock_data['info'], input_sources, is_indian_stock )
else: display(Markdown("❌ **Skipping Summary Dashboard due to prior errors.**"))


# %% [markdown]
# ## 10. Assumptions & Disclaimer (Reflects Dynamic/User Inputs & Fallbacks - Growth Logic Updated)

# %%
# Dynamically generate assumption strings based on final values (handle potential missing dicts)
# Get the actual source string for Rf (which now includes the default value if used)
rf_source_disp = input_sources.get('Rf', '?') # Already captures default info if applicable
rm_source_disp = input_sources.get('Rm','?')
beta_source_disp = input_sources.get('Beta','?')
c_spread_source_disp = input_sources.get('Credit Spread','N/A')
bench_source_disp = input_sources.get('Benchmarks','?')
growth_source_disp = fcff_results.get('growth_source', 'N/A') if fcff_results and isinstance(fcff_results, dict) else 'N/A'
tax_rate_source_disp = wacc_components.get('tax_rate_source', 'N/A') if wacc_components and isinstance(wacc_components, dict) else 'N/A'
equity_source_disp = wacc_components.get('equity_source','N/A') if wacc_components and isinstance(wacc_components, dict) else 'N/A'
weights_source_disp = wacc_components.get('weights_source','N/A') if wacc_components and isinstance(wacc_components, dict) else 'N/A'
terminal_growth_final = fcff_results.get('terminal_growth_rate', np.nan) if fcff_results and isinstance(fcff_results, dict) else np.nan
initial_growth_final = fcff_results.get('initial_growth_rate', np.nan) if fcff_results and isinstance(fcff_results, dict) else np.nan # Get final initial growth


# %%
# ----> UPDATED TEXT TO REFLECT NEW GROWTH LOGIC <----
display(Markdown(f"""
**Key Assumptions & Input Sources:**
*   **Financial Data:** Relies on data provided by Yahoo Finance (`yfinance`). Accuracy subject to provider. Missing data uses fallbacks or causes errors.
*   **Risk-Free Rate (Rf):** Attempted dynamic fetch, **confirmed/overridden by user.** (`Source: {rf_source_disp}` - Default suggestion: `{DEFAULT_RF_RATE_SUGGESTION:.1f}%`)
*   **Expected Market Return (Rm):** Based on **user input.** (`Source: {rm_source_disp}`)
*   **Market Risk Premium (MRP):** Derived from user `Rm` and final `Rf`.
*   **Beta (β):** Dynamically fetched if valid, otherwise **required user input.** (`Source: {beta_source_disp}`)
*   **Cost of Debt (Kd):** Implied if possible & within bounds (`{format_percent(MIN_COST_OF_DEBT)}` - `{format_percent(MAX_COST_OF_DEBT)}`). Otherwise, defaults to `Rf + Credit Spread`.
*   **Credit Spread:** Used only if default Kd triggered. Based on **user input** (default suggestion: `{format_percent(DEFAULT_CREDIT_SPREAD)}`). (`Source: {c_spread_source_disp}`)
*   **Growth Rates:**
    *   **Initial Growth:** `{format_percent(initial_growth_final)}` used for Year 1, determined via: (`Source: {growth_source_disp}`)
        1.  Calculate **3-Year Average Revenue Growth** (using `calculate_average_growth` which handles positive/negative values).
        2.  If step 1 fails (insufficient data/invalid calc), use **Default (`{format_percent(DEFAULT_GROWTH_RATE)}`)**.
        3.  Apply **Bounds:** The result from step 1 or 2 is then capped between **`{format_percent(MIN_INITIAL_GROWTH_FLOOR)}` (Floor)** and **`{format_percent(MAX_INITIAL_GROWTH_CEILING)}` (Ceiling)**.
    *   **Terminal Growth (g):** `{format_percent(terminal_growth_final)}` = min(GDP cap=`{format_percent(TERMINAL_GROWTH_RATE_CAP)}`, WACC*`0.67`, Final Initial Growth, Rf*`0.95`), floored at `{format_percent(MIN_GROWTH)}`, ensured `< WACC`.
*   **Tax Rate:** Effective rate `{tax_rate_source_disp}` (bounded `{format_percent(MIN_TAX_RATE)}` - `{format_percent(MAX_TAX_RATE)}`) or default (`{format_percent(DEFAULT_TAX_RATE)}`).
*   **Capital Structure:** Assumed stable using `{equity_source_disp}` and calculated debt. (`Weights Source: {weights_source_disp}`)
*   **Industry Benchmarks:** Uses **user input** or **generic fallbacks** if input left blank. (`Source: {bench_source_disp}`)
*   **Currency Formatting:** Uses ₹/Lakh/Crore for `.NS` tickers, standard K/M/B/T otherwise.

**Limitations:**
*   **Model Sensitivity:** DCF highly sensitive to inputs. **Sensitivity analysis (not included) crucial.**
*   **Input Quality:** Validity depends on user inputs and appropriateness of fallbacks.
*   **Historical Data Availability:** Financial statement data (needed for Revenue growth) is often limited to ~4 years via `yfinance`, which is sufficient for the 3-year calculation but relies on that limited data.
*   **Revenue Growth Calculation:** The average growth calculation handles fluctuations but may be skewed by outliers in the 3-year period. It is not a true CAGR.
*   **Dynamic Data Reliability:** Depends on `yfinance` accuracy/availability. Rf proxies need verification.
*   **Simplifications:** Linear decay, constant rates/structure. Ignores qualitative factors, macro shocks, sentiment.
*   **Fallback Accuracy:** Default growth rate (6.5%) and generic benchmarks may not be representative.
*   **Data Completeness:** Performance degrades with missing data. *Cross-verify key data.*

**Disclaimer:**
*For **educational and illustrative purposes only**. **NOT investment advice**. Models have limitations. Users MUST conduct own due diligence and consult a qualified financial advisor before investing. Creators NOT liable for actions based on this report.*
"""))


# %%




````

````python
# %% [markdown]
# # Dynamic Stock Financial Analysis & DCF Valuation
# **(Interactive Report - Dynamic/User Input Driven Assumptions - V11 - Refined GM Growth)**
#
# **Analyst:** Simulated AI Research Analyst
#
# **Objective:** To perform a comprehensive financial analysis and estimate the intrinsic value per share of a publicly traded company using the Discounted Cash Flow (DCF) methodology. Prioritizes dynamic fetching, falls back to user input, and uses configured defaults as a last resort. Includes fallback benchmarks for ratios and robust error handling/display. Adapts currency formatting for Indian stocks. **Initial growth uses a Geometric Mean of valid 3yr Rev/FCFF/Stock CAGRs (modified formulas for negatives), bounded & defaulted.**
#
# **Methodology:**
# 1.  **User Input & Dynamic Fetching:** Get ticker. Attempt dynamic Rf, require user confirmation/input for Rf, Rm, Beta (if needed), Credit Spread (if needed), and Ratio Benchmarks (with fallbacks). Detects Indian stocks for currency formatting.
# 2.  **Data Acquisition:** Use `yfinance` for statements, info, market data. Fetch extended price history (4y).
# 3.  **WACC Calculation:** Determine WACC using fetched/user inputs via CAPM. Handle errors.
# 4.  **FCFF Calculation & Projection:** Calculate historical FCFF. **Estimate initial growth by:**
#     *   Fetching last 3 years (4 data points) of Revenue, FCFF, and Adjusted Stock Price.
#     *   Checking validity of start (T-3) and end (T-0) points for each metric (start must be positive & finite).
#     *   Calculating 3-year CAGR *only* for metrics with valid data (using modified CAGR for negative end values).
#     *   Calculating the Geometric Mean of the valid calculated CAGRs (using modified GM for negative rates).
#     *   Falling back to default (6.5%) if GM fails or no valid CAGRs exist.
#     *   Bounding the result between 1.1% and 15%.
#     Project FCFF forward using linear decay to terminal growth. Handle errors.
# 5.  **Terminal Value:** Estimate value beyond the explicit forecast period.
# 6.  **Discounting & Valuation:** Discount cash flows and terminal value. Handle errors.
# 7.  **Ratio Analysis:** Calculate ratios and compare to user-input or fallback default benchmarks. Handles non-finite values.
# 8.  **Interactive Visualization:** Present findings using Plotly, including dynamic gauge ranges, robust display logic, and ₹/Lakh/Crore formatting for Indian stocks.

# %% [markdown]
# ## 1. Setup and Libraries

# %%
# Standard Libraries
import warnings
import sys
import numpy as np
import pandas as pd
import datetime
import math # For isnan, isinf checks

# Data Fetching
import yfinance as yf

# Visualization (Using Plotly)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

# Display Utilities
from IPython.display import display, Markdown, HTML

# Settings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pio.templates.default = "plotly_white" # Cleaner Plotly template

# %% [markdown]
# ## 2. Configuration and Helper Functions (Growth Helpers Updated)

# %%
# --- Model Configuration ---
DEFAULT_TICKER = "WIPRO.NS" # Default if user input fails
FORECAST_YEARS = 5
# Terminal Growth Rate CAP (Also acts as a default ceiling)
TERMINAL_GROWTH_RATE_CAP = 0.05 # Default/Cap set to 5%
# ----> DEFAULT INITIAL GROWTH (Used only if GM of CAGRs calc fails) <----
DEFAULT_GROWTH_RATE = 0.065  # Default set to 6.5%
# Default Risk-Free Rate suggestion (used if dynamic fetch fails)
DEFAULT_RF_RATE_SUGGESTION = 6.0 # Default set to 6.0% (as percent)

DEFAULT_CURRENCY_SYMBOL = "₹" # Changed default to Rupee, will be overridden by yfinance if possible
DEFAULT_TAX_RATE = 0.25     # General fallback tax rate if calculation fails
DEFAULT_CREDIT_SPREAD = 0.03 # Fallback spread over Rf for Cost of Debt
DEFAULT_MARKET_RISK_PREMIUM = 0.08 # Used only to suggest Rm to user

# --- Generic Fallback Benchmark Ratios (Used ONLY if User leaves input blank) ---
DEFAULT_BENCHMARK_RATIOS = {
    'P/E': 20.0,
    'P/B': 2.5,
    'EV/EBITDA': 12.0,
    'ROE': 15.0, # As percent value (e.g., 15 for 15%)
    'ROA': 8.0,  # As percent value
    'ROCE': 12.0,# As percent value
    'Asset Turnover': 0.8,
    'Current Ratio': 1.5,
    'Quick Ratio': 1.0
}

# --- Model Stability Bounds --- (Not primary financial assumptions, but guardrails)
MIN_WACC = 0.06
MAX_WACC = 0.25
MIN_TAX_RATE = 0.10
MAX_TAX_RATE = 0.40
MIN_GROWTH = 0.00 # Minimum growth floor (for terminal growth and projection steps)
# ----> Bounds for CALCULATED INITIAL GROWTH RATE <----
MIN_INITIAL_GROWTH_FLOOR = 0.011 # 1.1%
MAX_INITIAL_GROWTH_CEILING = 0.15  # 15.0%

MIN_COST_OF_DEBT = 0.03
MAX_COST_OF_DEBT = 0.15

# --- Ratio Keys Needed for Benchmarking ---
REQUIRED_BENCHMARK_RATIOS = list(DEFAULT_BENCHMARK_RATIOS.keys()) # Use keys from default dict

# --- Helper Functions ---
def print_keys(statement, name):
    """Prints available keys in a yfinance statement for debugging"""
    print(f"\n--- Available Keys in {name} ---")
    if isinstance(statement, pd.DataFrame):
        valid_indices = [idx for idx in statement.index if idx is not None]
        print(", ".join(map(str, valid_indices)))
    elif isinstance(statement, dict):
        print(", ".join(statement.keys()))
    print("------------------------------------\n")

def safe_get(data, key_list, statement_name="Statement", default=0.0):
    """
    Safely retrieves a numeric value from yfinance data (Series or Dict).
    Tries multiple keys, returns default if not found/invalid. Handles NaN/inf better.
    """
    if not isinstance(key_list, list):
        key_list = [key_list]

    value = None
    for key in key_list:
      current_value = None
      raw_val = None
      if isinstance(data, pd.Series):
          if key in data.index:
              try:
                  raw_val = data.loc[key]
                  # Check if raw_val is already numeric and finite before converting
                  if isinstance(raw_val, (int, float)) and np.isfinite(raw_val):
                      current_value = float(raw_val)
                  # Handle Series case (sometimes yfinance returns single value Series)
                  elif isinstance(raw_val, pd.Series):
                       if not raw_val.empty and pd.notna(raw_val.iloc[0]) and np.isfinite(raw_val.iloc[0]):
                            current_value = float(raw_val.iloc[0])
                  # Handle string conversion only if necessary and check finiteness
                  elif pd.notna(raw_val):
                      try_val = float(raw_val)
                      if np.isfinite(try_val):
                          current_value = try_val
              except (KeyError, TypeError, IndexError, ValueError):
                   pass # Ignore errors for this specific key/conversion
      elif isinstance(data, dict):
          raw_val = data.get(key)
          if raw_val is not None:
              try:
                  if isinstance(raw_val, str) and raw_val.lower() in ['n/a', 'none', '']:
                      pass # Treat as invalid
                  else:
                      temp_val = float(raw_val)
                      if np.isfinite(temp_val): # Check for NaN/inf after conversion
                           current_value = temp_val
              except (TypeError, ValueError):
                   pass

      if current_value is not None:
            value = current_value
            break # Found a valid value

    if value is None:
        value = default # Use the provided default if no valid value found
    return value

def format_currency(value, symbol="₹", precision=2, is_indian=False):
    """
    Formats number as currency string, handles None/NaN/inf.
    Uses Lakh/Crore for Indian stocks if is_indian=True.
    """
    if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
    if abs(value) > 1e18: return "Inf" # Handle extremely large numbers as Inf

    sign = "-" if value < 0 else ""
    abs_val = abs(value)

    if is_indian:
        # Use ₹ symbol regardless of fetched symbol for consistency
        symbol = "₹"
        if abs_val >= 1e7:  # 1 Crore (10 million)
            formatted_val = f"{abs_val / 1e7:.{precision}f} Cr"
        elif abs_val >= 1e5:  # 1 Lakh (100 thousand)
            formatted_val = f"{abs_val / 1e5:.{precision}f} Lakh"
        else:
            formatted_val = f"{abs_val:,.{precision}f}" # Add comma separators for smaller Rupee values
        return f"{sign}{symbol} {formatted_val}" # Add space after symbol
    else:
        # Standard international formatting (K, M, B, T)
        if abs_val >= 1e12: formatted_val = f"{abs_val / 1e12:.{precision}f}T"
        elif abs_val >= 1e9: formatted_val = f"{abs_val / 1e9:.{precision}f}B"
        elif abs_val >= 1e6: formatted_val = f"{abs_val / 1e6:.{precision}f}M"
        elif abs_val >= 1e3: formatted_val = f"{abs_val / 1e3:.{precision}f}K"
        else: formatted_val = f"{abs_val:.{precision}f}"
        return f"{sign}{symbol}{formatted_val}" # Standard symbol prefix

def format_percent(value, precision=1):
     """Formats number as percentage string, handles None/NaN/inf."""
     if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
     if abs(value) > 1e5: return "Inf" # Handle extreme % values

     return f"{value * 100:.{precision}f}%"

def get_validated_float_input(prompt_message, default_value=None, allow_blank=False, min_val=None, max_val=None):
    """Gets and validates user input for a float value, handles defaults, blanks, and bounds."""
    while True:
        prompt_with_default = prompt_message
        if default_value is not None:
            prompt_with_default += f" [Default: {default_value}]: "
        elif allow_blank:
             prompt_with_default += f" [Leave blank to skip/use fallback]: "
        else:
             prompt_with_default += ": "

        try:
            user_input_str = input(prompt_with_default).strip()

            if not user_input_str:
                if allow_blank: return None
                if default_value is not None:
                    try: return float(default_value)
                    except ValueError: raise ValueError(f"Internal Error: Invalid default value '{default_value}'.")
                raise ValueError("Input cannot be empty.")

            if user_input_str.endswith('%'): # Allow user to add % sign optionally
                user_input_str = user_input_str[:-1]

            value = float(user_input_str)

            if min_val is not None and value < min_val:
                 raise ValueError(f"Value must be >= {min_val}.")
            if max_val is not None and value > max_val:
                 raise ValueError(f"Value must be <= {max_val}.")

            return value
        except ValueError as e:
            print(f"❌ Invalid input: {e}. Please enter a valid number.")

def try_get_dynamic_rf(rf_ticker_symbol="^TNX"):
    """Attempts to fetch a dynamic risk-free rate (e.g., US 10Y Treasury)."""
    print(f"Attempting to fetch dynamic Risk-Free Rate ({rf_ticker_symbol})...")
    try:
        rf_ticker = yf.Ticker(rf_ticker_symbol)
        hist = rf_ticker.history(period="5d")
        if not hist.empty and 'Close' in hist.columns:
            last_close = hist['Close'].iloc[-1]
            if pd.notna(last_close):
                dynamic_rf = last_close / 100.0 # Assume Yahoo gives yield as %
                if 0.001 < dynamic_rf < 0.20: # Check bounds 0.1% to 20%
                    print(f"✅ Dynamically fetched Rf ({rf_ticker_symbol}): {format_percent(dynamic_rf)}")
                    return dynamic_rf
                else:
                     print(f"⚠️ Fetched Rf value ({format_percent(dynamic_rf)}) out of typical range. Ignoring.")
                     return None
            else:
                 print(f"⚠️ Fetched Rf data was invalid (NaN) for {rf_ticker_symbol}.")
                 return None
        else:
             print(f"⚠️ Could not fetch recent data for {rf_ticker_symbol}.")
             return None
    except Exception as e:
        print(f"❌ Error fetching dynamic Rf ({rf_ticker_symbol}): {e}")
        return None

# --- Growth Calculation Helpers (Updated) ---
# calculate_average_growth kept for reference
def calculate_average_growth(values_list):
    """Calculates average simple growth rate (NOT CAGR). Kept for reference."""
    growth_rates = []
    if len(values_list) < 2: return np.nan
    for i in range(len(values_list) - 1):
        val_t = values_list[i]; val_t_minus_1 = values_list[i+1]
        if pd.notna(val_t) and pd.notna(val_t_minus_1) and math.isfinite(val_t) and math.isfinite(val_t_minus_1):
            if abs(val_t_minus_1) > 1e-9:
                g = (val_t / val_t_minus_1) - 1
                if -10.0 < g < 10.0: growth_rates.append(g)
    return np.mean(growth_rates) if growth_rates else np.nan

def calculate_cagr_modified(start_val, end_val, num_years):
    """
    Calculates Compound Annual Growth Rate (CAGR).
    MODIFIED to handle cases where end_val is negative or zero, assuming start_val is positive.
    Returns NaN if start_val is non-positive, num_years <= 0, or values are invalid.
    Also returns NaN if result is extremely high/low (often indicates data issues).
    """
    if pd.isna(start_val) or pd.isna(end_val) or not math.isfinite(start_val) or not math.isfinite(end_val) or num_years <= 0:
        # print(f"CAGR Input Invalid: S={start_val}, E={end_val}, N={num_years}")
        return np.nan
    # CRITICAL: Start value must be positive for any meaningful CAGR interpretation.
    if start_val <= 1e-9:
        # print(f"CAGR Failed: Non-positive start value ({start_val})")
        return np.nan

    cagr = np.nan # Initialize cagr as nan
    try:
        # Handle non-positive end value
        if end_val <= 1e-9:
            # Method: Calculate CAGR based on absolute values, then make it negative.
            # This represents the annualized rate of decline towards zero or negative territory.
            end_val_abs = max(abs(end_val), 1e-9) # Use tiny positive value for calculation
            # Need to check if end_val_abs / start_val is positive before power operation
            ratio = end_val_abs / start_val
            if ratio > 0:
                 cagr_magnitude = ratio ** (1 / num_years) - 1
                 # Always negative if end value was non-positive relative to positive start
                 cagr = -cagr_magnitude
            else: # Should not happen if start_val > 0 and end_val_abs > 0
                 cagr = np.nan # Indicate failure
        else:
            # Standard CAGR calculation if both start and end are positive
            ratio = end_val / start_val
            if ratio > 0:
                 cagr = ratio ** (1 / num_years) - 1
            else: # Should not happen if start_val > 0 and end_val > 0
                 cagr = np.nan # Indicate failure

    except (ValueError, OverflowError, ZeroDivisionError) as e:
        # print(f"CAGR Numeric Error: S={start_val}, E={end_val}, N={num_years}, Error: {e}")
        cagr = np.nan # Ensure cagr is NaN on error

    # Final check for extremely large/small results which might indicate issues
    if pd.notna(cagr) and not (-0.9999 < cagr < 10.0): # e.g., reject >1000% or <-99.99% growth
        # print(f"CAGR Failed: Result out of reasonable bounds ({format_percent(cagr)})")
        cagr = np.nan # Treat extreme results as invalid

    # print(f"CAGR Result: S={start_val}, E={end_val}, N={num_years} -> CAGR={cagr}")
    return cagr


def geometric_mean_modified(rates):
    """
    Calculates the geometric mean of a list of rates (e.g., CAGRs).
    MODIFIED using an offset method to handle negative rates (down to nearly -100%).
    Falls back to arithmetic mean if any rate is <= -1.0 (-100%) or calculation fails.
    Returns NaN if no valid rates are provided.
    """
    # Filter out non-finite values first
    valid_rates = [r for r in rates if pd.notna(r) and math.isfinite(r)]
    if not valid_rates:
        # print("GM failed: No valid rates provided.")
        return np.nan
    if len(valid_rates) == 1:
        return valid_rates[0] # GM of one item is the item itself

    # Offset method: Add 1 to each rate. This factor MUST be > 0.
    offset = 1.0
    factors_offset = [r + offset for r in valid_rates]

    # Check if any rate was <= -1.0 (which makes the factor <= 0)
    if any(f <= 1e-9 for f in factors_offset): # Use a small threshold
         # print("⚠️ Geometric mean calculation failed due to rate <= -100%. Falling back to arithmetic mean.")
         # Calculate arithmetic mean only on the originally valid rates
         return np.mean(valid_rates)

    n = len(valid_rates)
    product = 1.0
    try:
        for factor in factors_offset:
             product *= factor
        # Calculate GM of offset rates. Product must be positive here.
        gm_offset = product ** (1/n)
        # Subtract the offset
        gm = gm_offset - offset
        # print(f"GM calculated: rates={valid_rates}, gm={gm}")

        # Add a check for extreme GM results
        if pd.notna(gm) and not (-0.9999 < gm < 10.0):
            # print(f"⚠️ Geometric mean result ({format_percent(gm)}) out of reasonable bounds. Falling back to arithmetic mean.")
            return np.mean(valid_rates)

        return gm
    except (ValueError, OverflowError, ZeroDivisionError) as e:
        # print(f"⚠️ Geometric mean calculation encountered numeric error. Falling back to arithmetic mean.")
        # Fallback to arithmetic mean on the originally valid rates
        return np.mean(valid_rates)

# --- End Growth Calculation Helpers ---


# %% [markdown]
# ## 3. User Input & Dynamic Fetching: Ticker, Rf, Rm, Benchmarks, Beta, Spread

# %%
input_sources = {}

# --- Get Ticker Symbol ---
user_ticker = input(f"Enter stock ticker (e.g., MSFT, AAPL, RELIANCE.NS) [Default: {DEFAULT_TICKER}]: ")
if not user_ticker: user_ticker = DEFAULT_TICKER
ticker_symbol = user_ticker.strip().upper()
display(Markdown(f"**Analyzing Ticker:** `{ticker_symbol}`"))

# ----> DETECT IF INDIAN STOCK for currency formatting <----
is_indian_stock = ticker_symbol.endswith(".NS")
if is_indian_stock:
    print("Info: Detected Indian stock (.NS). Using ₹/Lakh/Crore formatting.")

# --- Determine RF Ticker based on Stock Ticker ---
rf_proxy_ticker = "^TNX" # Default US 10Y
if is_indian_stock: rf_proxy_ticker = "^NS10"; print(f"Info: Using Indian Rf proxy: {rf_proxy_ticker}")
elif ticker_symbol.endswith(".L"): rf_proxy_ticker = "^FTLC"; print(f"Info: Using UK Rf proxy: {rf_proxy_ticker}")
# Add more country/exchange specific Rf proxies if needed

# --- Get Risk-Free Rate (Rf) ---
print("\n--- CAPM Inputs ---")
dynamic_rf_rate = try_get_dynamic_rf(rf_proxy_ticker)
rf_prompt = f"Enter Risk-Free Rate (Rf) as percentage (e.g., '4.5' for 4.5%). Fetched {rf_proxy_ticker}?"
# ----> Use NEW default suggestion if dynamic fetch fails <----
suggested_rf_val = dynamic_rf_rate * 100 if dynamic_rf_rate is not None else DEFAULT_RF_RATE_SUGGESTION

risk_free_rate_input_pct = get_validated_float_input(
    rf_prompt, default_value=f"{suggested_rf_val:.2f}", allow_blank=False, min_val=0.0, max_val=20.0
)
risk_free_rate_input = risk_free_rate_input_pct / 100.0

input_sources['Rf'] = 'Dynamic Fetch (User Confirmed/Adjusted)' if dynamic_rf_rate is not None else f'User Input (Default: {DEFAULT_RF_RATE_SUGGESTION:.1f}%)' # Updated source string slightly
display(Markdown(f"**Using Risk-Free Rate (Rf):** {format_percent(risk_free_rate_input)} (`Source: {input_sources['Rf']}`)"))

# --- Get Expected Market Return (Rm) ---
suggested_mrp = DEFAULT_MARKET_RISK_PREMIUM
suggested_rm = risk_free_rate_input + suggested_mrp
rm_prompt = f"Enter Expected Market Return (Rm) as percentage (e.g., '10'). Suggestion based on Rf + {format_percent(suggested_mrp)} MRP is {format_percent(suggested_rm)}"

while True:
    market_return_input_pct = get_validated_float_input(
        rm_prompt, default_value=f"{suggested_rm * 100:.1f}", allow_blank=False, min_val=-10.0, max_val=35.0
    )
    market_return_input = market_return_input_pct / 100.0
    if market_return_input >= risk_free_rate_input: break
    else: print(f"❌ Error: Rm ({format_percent(market_return_input)}) must be >= Rf ({format_percent(risk_free_rate_input)}). Re-enter.")

market_risk_premium_input = market_return_input - risk_free_rate_input
input_sources['Rm'] = 'User Input'
input_sources['MRP'] = 'Derived'
display(Markdown(f"**Using Expected Market Return (Rm):** {format_percent(market_return_input)} (`Source: {input_sources['Rm']}`)"))
display(Markdown(f"**↳ Implied Market Risk Premium (MRP):** {format_percent(market_risk_premium_input)} (`Source: {input_sources['MRP']}`)"))

# --- Beta (Placeholder) ---
beta_input = None; input_sources['Beta'] = 'Pending'
# --- Credit Spread (Placeholder) ---
user_credit_spread = None; input_sources['Credit Spread'] = 'Pending'

# --- Get Industry Benchmark Ratios (with fallback defaults) ---
print("\n--- Industry Ratio Benchmarks ---")
display(Markdown("Enter benchmarks or leave blank for generic fallbacks. *(Find industry averages online)*"))
user_benchmarks = {}; benchmark_sources = {}
print("Provide benchmarks (e.g., ROE: enter '15' for 15%; P/E: enter '20'):")
for ratio_key in REQUIRED_BENCHMARK_RATIOS:
    default_val = DEFAULT_BENCHMARK_RATIOS.get(ratio_key, None)
    prompt = f"  Enter benchmark for {ratio_key}"
    is_percent_type = ratio_key in ['ROE', 'ROA', 'ROCE']
    min_b, max_b = (-50, 100) if is_percent_type else (-100, 500)

    bench_val_input = get_validated_float_input(
        prompt, default_value=f"{default_val:.2f}" if default_val is not None else None,
        allow_blank=True, min_val=min_b, max_val=max_b
    )

    if bench_val_input is not None:
        user_benchmarks[ratio_key] = bench_val_input
        benchmark_sources[ratio_key] = 'User Input'
        display_val = f"{bench_val_input:.1f}%" if is_percent_type else f"{bench_val_input:.2f}"
        display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: User`)"))
    elif default_val is not None:
         user_benchmarks[ratio_key] = default_val # Store the default value if user skipped
         benchmark_sources[ratio_key] = 'Fallback Default'
         display_val = f"{default_val:.1f}%" if is_percent_type else f"{default_val:.2f}"
         display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: Default`)"))
    else:
        benchmark_sources[ratio_key] = 'Not Available'
        display(Markdown(f"  - **{ratio_key} Benchmark:** Not Provided / No Default"))
input_sources['Benchmarks'] = 'User Input or Default'

# %% [markdown]
# ## 4. Data Acquisition (Fetching Extended History)

# %%
stock_data = {}
try:
    stock = yf.Ticker(ticker_symbol)
    info = stock.info
    if not info or not info.get('regularMarketPrice'):
        print(f"Warning: Limited info for {ticker_symbol}. Trying history...")
        hist_check = stock.history(period="5d");
        if hist_check.empty: raise ValueError(f"Ticker '{ticker_symbol}' invalid/no data.")
        else: print(f"Warning: Proceeding with limited data."); info = {} if not isinstance(info, dict) else info
    stock_data['info'] = info

    # Standard statement fetching (usually 4 years)
    stock_data['financials'] = stock.financials
    stock_data['balance_sheet'] = stock.balance_sheet
    stock_data['cashflow'] = stock.cashflow
    if stock_data['financials'].empty or stock_data['balance_sheet'].empty or stock_data['cashflow'].empty:
       print(f"⚠️ Warning: Empty statements for {ticker_symbol}. DCF unreliable.")

    # Fetch recent history for charts/current price fallback
    stock_data['history'] = stock.history(period="2y")
    # ----> Fetch extended history (need at least 4 years for 3yr CAGR) <----
    stock_data['history_long'] = stock.history(period="4y") # 4 years should be sufficient for 3yr CAGR
    if stock_data['history_long'].empty or len(stock_data['history_long']) < 2: # Need at least 2 points
        print(f"⚠️ Warning: Could not fetch sufficient long history (4y) for {ticker_symbol}. Stock CAGR calc may fail.")


    company_name = info.get("longName", info.get("shortName", ticker_symbol))
    # Determine currency symbol, prioritizing Rupee for Indian stocks
    currency_code = info.get("currency", DEFAULT_CURRENCY_SYMBOL)
    if not isinstance(currency_code, str) or len(currency_code)>5 : currency_code=DEFAULT_CURRENCY_SYMBOL
    currency_symbol = "₹" if is_indian_stock else info.get("symbol", currency_code) # Prefer symbol if available, else code, force Rupee for Indian

    current_price = safe_get(info, ['currentPrice', 'regularMarketPrice', 'previousClose', 'lastPrice'], 'Info', default=np.nan)
    price_source = "yfinance Info"
    if not np.isfinite(current_price) and not stock_data['history'].empty:
        last_close_hist = stock_data['history']['Close'].iloc[-1]
        if pd.notna(last_close_hist): current_price = last_close_hist; price_source = "yfinance History"
    if not np.isfinite(current_price): print(f"⚠️ Warning: Could not determine current price."); current_price = np.nan

    shares_outstanding = safe_get(info, ['sharesOutstanding'], 'Info', default=0)

    beta_input = safe_get(info, ['beta'], 'Info', default=None)
    if beta_input is not None and np.isfinite(beta_input):
        input_sources['Beta'] = 'Dynamic Fetch'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))
    else:
         print(f"❌ ERROR: Beta not found dynamically or invalid.")
         beta_input = get_validated_float_input(f"Enter Beta for {ticker_symbol}:", allow_blank=False, min_val=-1.0, max_val=4.0)
         info['beta'] = beta_input; input_sources['Beta'] = 'User Input'
         display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))

    market_cap = safe_get(info, ['marketCap'], 'Info', default=0)
    mcap_source = "yfinance Info"
    if market_cap <= 0 and shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Calculated"; print(f"Info: Market Cap calculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}") # Pass flag
    elif market_cap <= 0: mcap_source = "Unavailable"; print(f"⚠️ Warning: Market Cap unavailable."); market_cap = np.nan

    sector = info.get("sector", "N/A"); industry = info.get("industry", "N/A")
    summary = info.get("longBusinessSummary", "No summary available.")

    display(Markdown(f"**Company:** {company_name}")); display(Markdown(f"**Sector:** {sector} | **Industry:** {industry}"))
    display(Markdown(f"**Currency:** {currency_code} ({currency_symbol})")); # Show both code and effective symbol
    # ----> Pass is_indian_stock flag to format_currency calls <----
    display(Markdown(f"**Current Market Price:** {format_currency(current_price, currency_symbol, is_indian=is_indian_stock)} (`Source: {price_source}`)"))
    display(Markdown(f"**Market Cap:** {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)} (`Source: {mcap_source}`)"))

except Exception as e: print(f"❌ CRITICAL ERROR fetching data: {e}"); sys.exit("Cannot proceed.")

if not isinstance(shares_outstanding, (int, float)) or shares_outstanding <= 0: sys.exit("Essential data missing: Shares Outstanding.")
if beta_input is None or not np.isfinite(beta_input): sys.exit("Essential data missing: Beta.")
if not np.isfinite(market_cap): print(f"⚠️ WARNING: Market Cap unavailable. WACC weights less accurate.")

# %% [markdown]
# ## 5. WACC Calculation

# %%
# WACC Function Definition (No changes here)
def calculate_wacc(stock_info_dict, financials_df, balance_sheet_df,
                   user_rf, user_mrp, user_beta, input_sources_dict,
                   currency_sym, is_indian): # Added currency params
    global user_credit_spread
    display(Markdown("### Calculating WACC..."))
    components = {}
    calculation_log = []
    calculation_log.append(f"Input Sources: Rf='{input_sources_dict.get('Rf')}', MRP='{input_sources_dict.get('MRP')}', Beta='{input_sources_dict.get('Beta')}'")

    beta_val = user_beta
    if beta_val <= 0: calculation_log.append(f"⚠️ Warning: Non-positive Beta ({beta_val:.2f}).")
    components['beta'] = beta_val; components['risk_free_rate'] = user_rf; components['market_risk_premium'] = user_mrp
    cost_of_equity = user_rf + beta_val * user_mrp
    components['cost_of_equity'] = cost_of_equity
    calculation_log.append(f"Inputs -> Rf: {format_percent(user_rf)}, Beta: {beta_val:.2f}, MRP: {format_percent(user_mrp)}")
    calculation_log.append(f"↳ Cost of Equity (Ke) [CAPM]: {format_percent(cost_of_equity)}")

    cost_of_debt_pre_tax = np.nan; tax_rate = np.nan; after_tax_cost_of_debt = np.nan
    total_debt = 0.0; debt_calculation_method = "N/A"; tax_rate_source = "N/A"

    if not balance_sheet_df.empty and not financials_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]; latest_fin = financials_df.iloc[:, 0]
        debt_keys = ['Total Debt', 'Long Term Debt', 'Long Term Debt And Capital Lease Obligation', 'Total Debt Non Current']
        long_term_debt = safe_get(latest_bs, debt_keys , 'Balance Sheet', default=0)
        current_debt_keys = ['Current Debt', 'Short Term Debt', 'Current Debt And Capital Lease Obligation', 'Short Long Term Debt', 'Total Debt Current']
        short_term_debt = safe_get(latest_bs, current_debt_keys , 'Balance Sheet', default=0)
        total_debt = long_term_debt + short_term_debt
        components['debt_value_used'] = total_debt
        interest_expense = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], 'Income Statement', default=0))

        implied_rate_calculated = False
        if total_debt > 1000 and interest_expense > 0:
             implied_rate = interest_expense / total_debt; implied_rate_calculated = True
             if MIN_COST_OF_DEBT <= implied_rate <= MAX_COST_OF_DEBT:
                  cost_of_debt_pre_tax = implied_rate; debt_calculation_method = "Implied";
                  # ----> Pass is_indian flag to format_currency calls <----
                  calculation_log.append(f"Interest Exp: {format_currency(interest_expense, symbol=currency_sym, is_indian=is_indian)}, Debt: {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
                  calculation_log.append(f"↳ Implied Cost of Debt (Kd pre-tax): {format_percent(cost_of_debt_pre_tax)}")
             else: calculation_log.append(f"⚠️ Implied Kd ({format_percent(implied_rate)}) out of bounds. Falling back.")
        elif total_debt > 1000: calculation_log.append(f"ℹ️ Found Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}) but no Interest Exp. Falling back.") # Pass flag

        if not debt_calculation_method.startswith("Implied"):
             if total_debt > 1000:
                 debt_calculation_method = "Default (Rf+Spread)"
                 if user_credit_spread is None:
                     spread_prompt = f"Implied Kd failed/NA. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
                     user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
                     user_credit_spread = user_credit_spread_pct / 100.0
                     input_sources_dict['Credit Spread'] = 'User Input (Default Kd)'
                 else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
                 cost_of_debt_pre_tax = user_rf + user_credit_spread
                 calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(cost_of_debt_pre_tax)}")
                 calculation_log.append(f"   (Spread Source: {input_sources_dict['Credit Spread']})")
             else:
                 debt_calculation_method = "Zero Debt"; cost_of_debt_pre_tax = 0.0; input_sources_dict['Credit Spread'] = 'N/A (Zero Debt)';
                 # ----> Pass is_indian flag to format_currency calls <----
                 calculation_log.append(f"ℹ️ Negligible Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}). Kd = 0%.")

        income_tax_expense = safe_get(latest_fin, ['Tax Provision', 'Income Tax Expense'], 'Income Statement', default=0)
        pretax_income = safe_get(latest_fin, ['Pretax Income', 'EBT', 'Earnings Before Tax'], 'Income Statement', default=0)
        if pretax_income <= 0 and income_tax_expense > 0: # Handle negative pretax income if tax was paid
             ebit_val = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=0); interest_exp_for_ebt = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], default=0))
             pretax_income_from_ebit = ebit_val - interest_exp_for_ebt
             if pretax_income_from_ebit > 0: pretax_income = pretax_income_from_ebit # Use calculated if positive
        if pretax_income > 0 and income_tax_expense >= 0:
            calculated_tax_rate = income_tax_expense / pretax_income; tax_rate = min(max(calculated_tax_rate, MIN_TAX_RATE), MAX_TAX_RATE); tax_rate_source = "Derived & Bounded"
            if abs(tax_rate - calculated_tax_rate) > 0.001: tax_rate_source = f"Derived ({format_percent(calculated_tax_rate)}), Bounded"
            # ----> Pass is_indian flag to format_currency calls <----
            calculation_log.append(f"Tax Exp: {format_currency(income_tax_expense, symbol=currency_sym, is_indian=is_indian)}, Pretax Inc: {format_currency(pretax_income, symbol=currency_sym, is_indian=is_indian)}")
            calculation_log.append(f"↳ Effective Tax Rate: {format_percent(tax_rate)} ({tax_rate_source})")
        else: tax_rate = DEFAULT_TAX_RATE; tax_rate_source = "Default"; calculation_log.append(f"⚠️ Cannot derive tax rate. Using default: {format_percent(tax_rate)}")
    else: # Financial statements empty
         calculation_log.append("⚠️ Financial statements empty. Using defaults."); debt_calculation_method = "Default (No Fin.)"; tax_rate_source = "Default (No Fin.)"
         if user_credit_spread is None:
              spread_prompt = f"No financials. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
              user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
              user_credit_spread = user_credit_spread_pct / 100.0; input_sources_dict['Credit Spread'] = 'User Input (No Fin.)'
         else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
         cost_of_debt_pre_tax = user_rf + user_credit_spread; tax_rate = DEFAULT_TAX_RATE; total_debt = 0; components['debt_value_used'] = 0
         calculation_log.append(f"Using Default Kd: {format_percent(cost_of_debt_pre_tax)} (Rf + Spread: {input_sources_dict['Credit Spread']})")
         calculation_log.append(f"Using Default Tax Rate: {format_percent(tax_rate)}")

    components['cost_of_debt'] = cost_of_debt_pre_tax if np.isfinite(cost_of_debt_pre_tax) else 0.0
    components['debt_calc_method'] = debt_calculation_method
    components['tax_rate'] = tax_rate if np.isfinite(tax_rate) else DEFAULT_TAX_RATE
    components['tax_rate_source'] = tax_rate_source
    if np.isfinite(cost_of_debt_pre_tax) and np.isfinite(tax_rate): after_tax_cost_of_debt = cost_of_debt_pre_tax * (1 - tax_rate)
    else: after_tax_cost_of_debt = components['cost_of_debt'] * (1 - DEFAULT_TAX_RATE) # Use default if calculation failed
    components['after_tax_cost_of_debt'] = after_tax_cost_of_debt
    if components['cost_of_debt'] > 0 : calculation_log.append(f"↳ Cost of Debt (After-tax): {format_percent(after_tax_cost_of_debt)}")

    market_cap_val = safe_get(stock_info_dict, ['marketCap'], 'Info', default=np.nan); equity_source = "Market Cap" if np.isfinite(market_cap_val) and market_cap_val > 0 else "N/A"
    equity_value_for_wacc = np.nan
    if np.isfinite(market_cap_val) and market_cap_val > 0: equity_value_for_wacc = market_cap_val
    else:
       calculation_log.append("⚠️ Market Cap unavailable/zero."); equity_source = "Book Equity Fallback"
       if not balance_sheet_df.empty:
           equity_keys = ['Total Stockholder Equity', 'Stockholders Equity', 'Total Equity Gross Minority Interest', 'Common Stock Equity']
           book_equity_val = safe_get(balance_sheet_df.iloc[:, 0], equity_keys, 'Balance Sheet', default=np.nan)
           if np.isfinite(book_equity_val) and book_equity_val > 1.0:
              equity_value_for_wacc = book_equity_val; calculation_log.append(f"Using Book Equity ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}) as proxy.") # Pass flag
           else: equity_source = "Placeholder (Book Invalid)"
       else: equity_source = "Placeholder (No BS)"
       if not np.isfinite(equity_value_for_wacc): equity_value_for_wacc = 1.0; calculation_log.append("⚠️ Market Cap & Book Equity unavailable. Using placeholder Equity Value (1.0). Weights inaccurate.")
    components['equity_value_used'] = equity_value_for_wacc; components['equity_source'] = equity_source

    equity_weight = np.nan; debt_weight = np.nan; weights_source = "N/A"; total_capital = equity_value_for_wacc + total_debt
    if np.isfinite(total_capital) and total_capital > 1.0:
        equity_weight = equity_value_for_wacc / total_capital; debt_weight = total_debt / total_capital; weights_source = "Calculated"
        total_weight = equity_weight + debt_weight # Normalize if calculation slightly off
        if abs(total_weight - 1.0) > 1e-6: equity_weight /= total_weight; debt_weight /= total_weight
    else: calculation_log.append(f"❌ ERROR: Total Capital ({format_currency(total_capital, symbol=currency_sym, is_indian=is_indian)}) invalid. Using default weights (70/30)."); equity_weight = 0.7; debt_weight = 0.3; weights_source = "Default" # Pass flag
    components['equity_weight'] = equity_weight; components['debt_weight'] = debt_weight; components['weights_source'] = weights_source
    # ----> Pass is_indian flag to format_currency calls <----
    calculation_log.append(f"Equity Value Used ({components['equity_source']}): {format_currency(components['equity_value_used'], symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"Debt Value Used ({components['debt_calc_method']}): {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"↳ Weights E/V: {format_percent(equity_weight)}, D/V: {format_percent(debt_weight)} (Source: {weights_source})")

    wacc = np.nan
    if all(np.isfinite(v) for v in [equity_weight, cost_of_equity, debt_weight, after_tax_cost_of_debt]):
        wacc = (equity_weight * cost_of_equity) + (debt_weight * after_tax_cost_of_debt); original_wacc = wacc; wacc = min(max(wacc, MIN_WACC), MAX_WACC)
        bound_note = ""
        if abs(wacc - original_wacc) > 0.0001 : bound_note = f"(Bounded from {format_percent(original_wacc)})"; calculation_log.append(f"⚠️ WACC ({format_percent(original_wacc)}) bounded.")
        calculation_log.append(f"▶️ Calculated WACC: {format_percent(wacc)} {bound_note}")
    else: calculation_log.append(f"❌ ERROR: Cannot calculate final WACC due to invalid components."); wacc = np.nan
    components['final_wacc'] = wacc

    display(Markdown("```\n" + "\n".join(calculation_log) + "\n```"))

    if np.isfinite(wacc) and wacc > 0 and np.isfinite(equity_weight) and np.isfinite(debt_weight):
        equity_contrib = equity_weight * cost_of_equity; debt_contrib = debt_weight * after_tax_cost_of_debt
        # Check if components reasonably sum up (handle potential float issues)
        if np.isfinite(equity_contrib) and np.isfinite(debt_contrib) and abs(equity_contrib + debt_contrib - wacc) < 0.01:
           labels = ['Cost of Equity Contribution', 'Cost of Debt Contribution']; values = [equity_contrib, debt_contrib]
           fig_wacc_donut = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4, marker_colors=px.colors.qualitative.Pastel, hovertemplate='<b>%{label}</b><br>Contribution: %{value:.2%}<extra></extra>', textinfo='percent', textfont_size=12, sort=False)])
           fig_wacc_donut.update_layout(title_text=f'WACC Composition (Total: {format_percent(wacc)})', title_x=0.5, showlegend=True, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), margin=dict(l=20, r=20, t=60, b=50), height=350)
           fig_wacc_donut.show()
        else: display(Markdown(f"*(WACC component contributions invalid/don't sum. Cannot generate chart.)*"))
    elif not np.isfinite(wacc): display(Markdown(f"*(WACC calculation failed. Cannot generate chart)*"))
    else: display(Markdown(f"*(WACC zero/negative or weights invalid. Cannot generate chart)*"))

    return wacc, components

# Execute WACC Calculation
# ----> Pass currency symbol and flag <----
discount_rate, wacc_components = calculate_wacc(
    stock_data['info'], stock_data['financials'], stock_data['balance_sheet'],
    risk_free_rate_input, market_risk_premium_input, beta_input, input_sources,
    currency_symbol, is_indian_stock
)

# %% [markdown]
# ## 6. FCFF Calculation & Projection (Initial Growth: Refined GM of valid 3yr CAGRs)

# %%
# FCFF Function Definition (NEW initial growth logic - V11)
def calculate_and_project_fcff(financials_df, balance_sheet_df, cashflow_df, history_long_df, # History needed for stock CAGR
                               wacc_comps, disc_rate, forecast_yrs,
                               final_user_rf, default_init_growth, # This is 6.5% default
                               currency_sym, is_indian): # Added currency params
    display(Markdown("### Calculating & Projecting FCFF..."))
    projection_log = []
    projection_results = {
        'historical_fcffs': {}, # Store recent historical FCFFs
        'historical_revenues': {}, # Store recent historical revenues
        'current_fcff': np.nan, 'projected_fcffs': [], 'growth_rates': [],
        'pv_fcffs': [], 'cumulative_pv_fcff': np.nan, 'terminal_growth_rate': np.nan,
        'terminal_value': np.nan, 'pv_terminal_value': np.nan, 'error': None,
        'initial_growth_rate': np.nan, 'growth_source': 'N/A'
    }

    if financials_df.empty or balance_sheet_df.empty or cashflow_df.empty:
         projection_results['error'] = "Missing Financials/Balance Sheet/Cash Flow."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results
    if not np.isfinite(disc_rate):
         projection_results['error'] = "Invalid WACC provided."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results

    # --- Define Helper to calculate FCFF for a specific year column ---
    def get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, tax_rt):
        if not isinstance(fin_col, pd.Series) or not isinstance(bs_col, pd.Series) or not isinstance(cf_col, pd.Series): return np.nan, "Missing Statement Column"
        ebit = safe_get(fin_col, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
        depreciation_amortization = safe_get(cf_col, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
        capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
        capital_expenditures = abs(safe_get(cf_col, capex_keys, default=0))
        wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]
        change_in_wc = safe_get(cf_col, wc_keys, default=None) ; wc_src = "CF"
        if change_in_wc is None or not np.isfinite(change_in_wc):
            wc_src = "BS Calc"
            if isinstance(bs_prev_col, pd.Series):
                 curr_assets = safe_get(bs_col, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(bs_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                 prev_assets = safe_get(bs_prev_col, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(bs_prev_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                 if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val): change_in_wc = curr_wc_val - prev_wc_val
                 else: change_in_wc = 0; wc_src = "BS Fail"
            else: change_in_wc = 0; wc_src = "No Prev BS"
        if not np.isfinite(change_in_wc): change_in_wc = 0
        fcff_val = np.nan; calc_status = "OK"
        if np.isfinite(ebit) and np.isfinite(tax_rt):
            nopat_val = ebit * (1 - tax_rt)
            if np.isfinite(depreciation_amortization) and np.isfinite(capital_expenditures) and np.isfinite(change_in_wc): fcff_val = nopat_val + depreciation_amortization - capital_expenditures - change_in_wc
            else: calc_status = "Invalid Depr/CapEx/WC"
        else: calc_status = "Invalid EBIT/Tax"
        return fcff_val, calc_status
    # --- End FCFF Helper ---

    # --- Calculate Historical FCFF & Revenue (need T-0 and T-3 for CAGR) ---
    num_years_cagr = 3
    max_hist_years = num_years_cagr + 1 # Need 4 data points for 3yr CAGR
    num_statement_years = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    historical_fcffs_list = [] # Newest first (index 0 = T-0)
    historical_revenues_list = [] # Newest first
    effective_tax_rate = wacc_comps.get('tax_rate', DEFAULT_TAX_RATE)

    projection_log.append(f"Calculating historical FCFF & Revenue (up to {max_hist_years} points)...")
    years_to_calc = min(num_statement_years, max_hist_years)
    for i in range(years_to_calc):
        fin_col = financials_df.iloc[:, i]; bs_col = balance_sheet_df.iloc[:, i]; cf_col = cashflow_df.iloc[:, i]
        bs_prev_col = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None
        fcff_val, status = get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, effective_tax_rate)
        historical_fcffs_list.append(fcff_val if status == "OK" else np.nan)
        rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
        revenue_val = safe_get(fin_col, rev_keys, default=np.nan)
        historical_revenues_list.append(revenue_val)

    projection_results['historical_fcffs'] = {f"T-{i}": v for i, v in enumerate(historical_fcffs_list)}
    projection_results['historical_revenues'] = {f"T-{i}": v for i, v in enumerate(historical_revenues_list)}
    fcff_current = historical_fcffs_list[0] if historical_fcffs_list and np.isfinite(historical_fcffs_list[0]) else np.nan
    projection_results['current_fcff'] = fcff_current

    # --- Display Calculation Log for Current FCFF (if calculated) ---
    if np.isfinite(fcff_current):
       latest_fin = financials_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]
       ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
       nopat = ebit * (1-effective_tax_rate) if np.isfinite(ebit) else np.nan
       depreciation_amortization = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
       capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]; capital_expenditures = abs(safe_get(latest_cf, capex_keys, default=0))
       wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]; change_in_wc = safe_get(latest_cf, wc_keys, default=None); wc_source = "Direct from Cash Flow"
       if change_in_wc is None or not np.isfinite(change_in_wc):
            if balance_sheet_df.shape[1] >= 2:
                prev_bs = balance_sheet_df.iloc[:, 1]
                curr_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0); curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                prev_assets = safe_get(prev_bs, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(prev_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0); prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val): change_in_wc = curr_wc_val - prev_wc_val; wc_source = "Calculated from BS"
                else: change_in_wc = 0; wc_source = "Calc Failed (Invalid BS)"
            else: change_in_wc = 0; wc_source = "Default (No Prev BS)"
       if not np.isfinite(change_in_wc): change_in_wc = 0
       projection_log.append(f"EBIT (T-0): {format_currency(ebit, currency_sym, is_indian=is_indian)}, Tax Rate ({wacc_comps.get('tax_rate_source','?')}) {format_percent(effective_tax_rate)} -> NOPAT: {format_currency(nopat, currency_sym, is_indian=is_indian)}")
       projection_log.append(f"  + Dep&Amort: {format_currency(depreciation_amortization, currency_sym, is_indian=is_indian)}")
       projection_log.append(f"  - CapEx: {format_currency(capital_expenditures, currency_sym, is_indian=is_indian)}")
       projection_log.append(f"  - Change WC ({wc_source}): {format_currency(change_in_wc, currency_sym, is_indian=is_indian)}")
       projection_log.append(f"▶️ Historical FCFF (T-0): {format_currency(fcff_current, currency_sym, is_indian=is_indian)}")
    else:
        projection_log.append("⚠️ Could not calculate current historical FCFF (T-0).")

    # Exit if T-0 FCFF is invalid, as it's needed for projection start.
    if not np.isfinite(fcff_current):
        projection_results['error'] = "Invalid historical FCFF (T-0). Cannot proceed.";
        display(Markdown(f"❌ ERROR: {projection_results['error']}")); display(Markdown("```\n" + "\n".join(projection_log) + "\n```")); return projection_results

    # --- REFINED Initial Growth Rate Calculation (GM of valid 3yr CAGRs) ---
    projection_log.append("--- Determining Initial Growth Rate (Geometric Mean of valid 3yr CAGRs) ---")
    initial_growth_rate = np.nan # Start with NaN
    growth_source = "N/A"
    calculated_cagrs = {} # Store individual CAGRs {Name: value}
    valid_cagr_list = [] # Store only the valid CAGRs for GM calculation

    # 1. Calculate 3-Year Revenue CAGR (Conditional)
    rev_cagr = np.nan
    projection_log.append(f"Calculating Revenue {num_years_cagr}yr CAGR...")
    if len(historical_revenues_list) >= max_hist_years:
        rev_end = historical_revenues_list[0]           # T-0
        rev_start = historical_revenues_list[num_years_cagr] # T-3
        # Check data validity BEFORE calculating CAGR
        if pd.notna(rev_end) and math.isfinite(rev_end) and pd.notna(rev_start) and math.isfinite(rev_start) and rev_start > 1e-9:
            rev_cagr = calculate_cagr_modified(rev_start, rev_end, num_years_cagr)
            if pd.notna(rev_cagr) and math.isfinite(rev_cagr):
                 calculated_cagrs['Revenue'] = rev_cagr
                 valid_cagr_list.append(rev_cagr)
                 projection_log.append(f"  ✅ Revenue CAGR: {format_percent(rev_cagr)} (Start: {format_currency(rev_start, '', 0, is_indian)}, End: {format_currency(rev_end, '', 0, is_indian)})")
            else:
                 projection_log.append(f"  ❌ Revenue CAGR Calculation Failed (Result Invalid). (Start: {format_currency(rev_start, '', 0, is_indian)}, End: {format_currency(rev_end, '', 0, is_indian)})")
        else:
            projection_log.append(f"  ❌ Revenue CAGR Skipped (Invalid/Non-positive Start Data). (Start: {format_currency(rev_start, '', 0, is_indian)}, End: {format_currency(rev_end, '', 0, is_indian)})")
    else:
        projection_log.append(f"  ❌ Revenue CAGR Skipped (Insufficient data: {len(historical_revenues_list)} points)")

    # 2. Calculate 3-Year FCFF CAGR (Conditional)
    fcff_cagr = np.nan
    projection_log.append(f"Calculating FCFF {num_years_cagr}yr CAGR...")
    if len(historical_fcffs_list) >= max_hist_years:
        fcff_end = historical_fcffs_list[0]           # T-0
        fcff_start = historical_fcffs_list[num_years_cagr] # T-3
        # Check data validity BEFORE calculating CAGR
        if pd.notna(fcff_end) and math.isfinite(fcff_end) and pd.notna(fcff_start) and math.isfinite(fcff_start) and fcff_start > 1e-9:
            fcff_cagr = calculate_cagr_modified(fcff_start, fcff_end, num_years_cagr)
            if pd.notna(fcff_cagr) and math.isfinite(fcff_cagr):
                 calculated_cagrs['FCFF'] = fcff_cagr
                 valid_cagr_list.append(fcff_cagr)
                 projection_log.append(f"  ✅ FCFF CAGR:    {format_percent(fcff_cagr)} (Start: {format_currency(fcff_start, '', 0, is_indian)}, End: {format_currency(fcff_end, '', 0, is_indian)})")
            else:
                 projection_log.append(f"  ❌ FCFF CAGR Calculation Failed (Result Invalid). (Start: {format_currency(fcff_start, '', 0, is_indian)}, End: {format_currency(fcff_end, '', 0, is_indian)})")
        else:
            projection_log.append(f"  ❌ FCFF CAGR Skipped (Invalid/Non-positive Start Data). (Start: {format_currency(fcff_start, '', 0, is_indian)}, End: {format_currency(fcff_end, '', 0, is_indian)})")
    else:
        projection_log.append(f"  ❌ FCFF CAGR Skipped (Insufficient data: {len(historical_fcffs_list)} points)")

    # 3. Calculate 3-Year Stock Price CAGR (Conditional)
    stock_cagr = np.nan
    projection_log.append(f"Calculating Stock Price {num_years_cagr}yr CAGR...")
    if history_long_df is not None and not history_long_df.empty and 'Adj Close' in history_long_df.columns and len(history_long_df) >= 2:
         try:
             end_date = history_long_df.index[-1]
             start_date_target = end_date - pd.DateOffset(years=num_years_cagr)
             end_idx = len(history_long_df) - 1
             start_idx = history_long_df.index.searchsorted(start_date_target, side='left')

             if start_idx < end_idx: # Ensure start is before end
                start_price = history_long_df['Adj Close'].iloc[start_idx]
                end_price = history_long_df['Adj Close'].iloc[end_idx]
                start_actual_date = history_long_df.index[start_idx]
                end_actual_date = history_long_df.index[end_idx]
                actual_years = (end_actual_date - start_actual_date).days / 365.25

                # Check data validity BEFORE calculating CAGR
                if pd.notna(end_price) and math.isfinite(end_price) and pd.notna(start_price) and math.isfinite(start_price) and start_price > 1e-9 and actual_years >= (num_years_cagr - 0.5) and actual_years > 0:
                    stock_cagr = calculate_cagr_modified(start_price, end_price, actual_years)
                    if pd.notna(stock_cagr) and math.isfinite(stock_cagr):
                        calculated_cagrs['Stock'] = stock_cagr
                        valid_cagr_list.append(stock_cagr)
                        projection_log.append(f"  ✅ Stock CAGR:   {format_percent(stock_cagr)} (Start: {start_price:.2f} on {start_actual_date.date()}, End: {end_price:.2f} on {end_actual_date.date()}, ~{actual_years:.2f} yrs)")
                    else:
                        projection_log.append(f"  ❌ Stock CAGR Calculation Failed (Result Invalid). (Start: {start_price:.2f}, End: {end_price:.2f}, Years: {actual_years:.2f})")
                else:
                    projection_log.append(f"  ❌ Stock CAGR Skipped (Invalid/Non-positive Start Data or Period Invalid). (Start: {start_price:.2f}, End: {end_price:.2f}, Years: {actual_years:.2f})")
             else:
                  projection_log.append(f"  ❌ Stock CAGR Skipped (Could not find distinct start/end points in history: StartIdx={start_idx}, EndIdx={end_idx})")
         except Exception as e:
              projection_log.append(f"  ❌ Stock CAGR Skipped (Error during lookup/calculation: {e})")
    else:
         projection_log.append(f"  ❌ Stock CAGR Skipped (Insufficient/invalid history data)")

    # 4. Calculate Geometric Mean of the valid CAGRs
    if valid_cagr_list:
        gm_growth = geometric_mean_modified(valid_cagr_list)
        if pd.notna(gm_growth) and math.isfinite(gm_growth):
            initial_growth_rate = gm_growth
            growth_source = f"GM of {len(valid_cagr_list)} valid CAGR(s)"
            projection_log.append(f"  ▶️ Calculated Geometric Mean = {format_percent(initial_growth_rate)}")
            included_cagrs = [f"{k}:{format_percent(v)}" for k, v in calculated_cagrs.items() if pd.notna(v) and math.isfinite(v)] # Rebuild list from dict
            projection_log.append(f"     Based on: [{', '.join(included_cagrs)}]")
        else:
            projection_log.append(f"  ⚠️ Geometric Mean calculation failed or result invalid. Falling back to default.")
            initial_growth_rate = default_init_growth
            growth_source = f"Default ({format_percent(default_init_growth)}) [GM Fail]"
            projection_log.append(f"     CAGRs calculated but GM failed: {valid_cagr_list}") # Log the rates that caused GM failure
    else:
        # If NO valid CAGRs were calculated
        projection_log.append(f"  ⚠️ No valid 3yr CAGRs could be calculated. Falling back to default.")
        initial_growth_rate = default_init_growth
        growth_source = f"Default ({format_percent(default_init_growth)}) [No Valid CAGRs]"

    # 5. Apply Floor (1.1%) and Ceiling (15%)
    unbounded_rate = initial_growth_rate # Store original for comparison
    initial_growth_rate = max(initial_growth_rate, MIN_INITIAL_GROWTH_FLOOR) # Apply 1.1% floor
    initial_growth_rate = min(initial_growth_rate, MAX_INITIAL_GROWTH_CEILING) # Apply 15% ceiling

    if abs(initial_growth_rate - unbounded_rate) > 1e-6: # Check if bounding changed the value
        growth_source += " (Bounded)"
        projection_log.append(f"  Applied Bounds: Final Initial Growth Rate = {format_percent(initial_growth_rate)} (Floor: {format_percent(MIN_INITIAL_GROWTH_FLOOR)}, Ceiling: {format_percent(MAX_INITIAL_GROWTH_CEILING)})")
    else:
         projection_log.append(f"  Final Initial Growth Rate = {format_percent(initial_growth_rate)}")

    projection_results['initial_growth_rate'] = initial_growth_rate
    projection_results['growth_source'] = growth_source

    # --- Terminal Growth Rate Calculation (Remains the same conservative approach) ---
    # Use the final, bounded initial_growth_rate here
    g_cap_initial_valid = initial_growth_rate if pd.notna(initial_growth_rate) and math.isfinite(initial_growth_rate) else TERMINAL_GROWTH_RATE_CAP # Fallback if initial is invalid

    g_cap_config = TERMINAL_GROWTH_RATE_CAP; g_cap_wacc = disc_rate * 0.67; g_cap_rf = final_user_rf * 0.95 # Use final RF
    projection_log.append(f"--- Determining Terminal Growth Rate ---")
    projection_log.append(f"Terminal Growth Caps Considered: Config={format_percent(g_cap_config)}, WACC*0.67={format_percent(g_cap_wacc)}, Initial(Final)={format_percent(g_cap_initial_valid)}, Rf*0.95={format_percent(g_cap_rf)}")
    terminal_growth_rate = min(g_cap_config, g_cap_wacc, g_cap_initial_valid, g_cap_rf); projection_log.append(f"↳ Minimum of Caps: {format_percent(terminal_growth_rate)}")
    original_g_before_floor = terminal_growth_rate
    terminal_growth_rate = max(MIN_GROWTH, terminal_growth_rate) # Apply floor (MIN_GROWTH=0.00 usually)
    if abs(terminal_growth_rate - original_g_before_floor) > 1e-6 : projection_log.append(f"↳ Adjusted by Floor ({format_percent(MIN_GROWTH)}): {format_percent(terminal_growth_rate)}")
    if terminal_growth_rate >= disc_rate: original_g = terminal_growth_rate; terminal_growth_rate = disc_rate - 0.005; projection_log.append(f"⚠️ Terminal Growth ({format_percent(original_g)}) >= WACC ({format_percent(disc_rate)}). Reduced to {format_percent(terminal_growth_rate)}.")
    projection_log.append(f"▶️ Using Final Terminal Growth Rate (g): {format_percent(terminal_growth_rate)}"); projection_results['terminal_growth_rate'] = terminal_growth_rate

    # --- Project FCFFs using Linear Decay from initial_growth_rate -> terminal_growth_rate ---
    projected_fcffs = []; growth_rates_applied = []; pv_fcffs = []; cumulative_pv_fcff = 0.0
    current_fcff_for_proj = fcff_current # Start projection base from T-0 FCFF

    # Ensure effective initial growth is valid (should be after default/bounding)
    if pd.isna(initial_growth_rate) or not math.isfinite(initial_growth_rate):
        projection_log.append(f"⚠️ FATAL: Invalid final initial growth rate ({initial_growth_rate}) despite checks. Using 0% for projection start.")
        effective_initial_growth = 0.0
    else:
        effective_initial_growth = initial_growth_rate # Use the final bounded rate

    effective_terminal_growth = terminal_growth_rate

    # Handle negative starting FCFF carefully
    if current_fcff_for_proj <= 0:
        projection_log.append("⚠️ Starting FCFF (T-0) <= 0. Applying minimal growth.")
        effective_initial_growth = min(effective_initial_growth, 0.01); # Minimal positive initial growth
        effective_terminal_growth = min(terminal_growth_rate, 0.00) # Minimal non-negative terminal

    projection_log.append("--- FCFF Projections ---"); projection_log.append(f"{'Year':<5} {'Growth':<8} {'FCFF':<18} {'PV Factor':<10} {'PV FCFF':<18}"); projection_log.append("-" * 60)
    last_valid_fcff = current_fcff_for_proj # Use the starting FCFF (T-0) for year 1 growth calculation

    for year in range(1, forecast_yrs + 1):
        if forecast_yrs > 0: decay_factor = max(0, (forecast_yrs - year)) / forecast_yrs
        else: decay_factor = 0
        growth_rate = effective_terminal_growth + (effective_initial_growth - effective_terminal_growth) * decay_factor
        growth_rate = max(MIN_GROWTH, growth_rate) # Apply general floor (0%)

        if current_fcff_for_proj <= 0: growth_rate = max(-0.10, min(growth_rate, 0.05)) # Moderate if start neg

        growth_rates_applied.append(growth_rate)
        fcff_forecast = last_valid_fcff * (1 + growth_rate)
        if not np.isfinite(fcff_forecast): fcff_forecast = 0
        projected_fcffs.append(fcff_forecast)
        last_valid_fcff = fcff_forecast # Update last_valid_fcff

        pv_factor = 1 / ((1 + disc_rate) ** year); pv_fcff = fcff_forecast * pv_factor
        if not np.isfinite(pv_fcff): pv_fcff = 0
        pv_fcffs.append(pv_fcff); cumulative_pv_fcff += pv_fcff
        projection_log.append(f"{year:<5} {format_percent(growth_rate):<8} {format_currency(fcff_forecast, '', 0, is_indian=is_indian):<18} {pv_factor:<10.4f} {format_currency(pv_fcff, '', 0, is_indian=is_indian):<18}")

    projection_results['projected_fcffs'] = projected_fcffs; projection_results['growth_rates'] = growth_rates_applied; projection_results['pv_fcffs'] = pv_fcffs
    projection_results['cumulative_pv_fcff'] = cumulative_pv_fcff if np.isfinite(cumulative_pv_fcff) else 0.0
    projection_log.append("-" * 60); projection_log.append(f"Sum of PV(Forecasted FCFFs): {format_currency(cumulative_pv_fcff, symbol=currency_sym, is_indian=is_indian)}")

    # Calculate Terminal Value (Remains the same)
    terminal_value = np.nan; pv_terminal_value = np.nan
    final_forecast_fcff = projected_fcffs[-1] if projected_fcffs and np.isfinite(projected_fcffs[-1]) else 0

    if (disc_rate - terminal_growth_rate) <= 1e-6 or not np.isfinite(terminal_growth_rate):
         projection_log.append(f"❌ ERROR: WACC ({format_percent(disc_rate)}) <= Terminal Growth ({format_percent(terminal_growth_rate)}) or g invalid. TV calc failed."); projection_results['error'] = "WACC <= g or g invalid. TV Calc failed."
    elif not np.isfinite(final_forecast_fcff):
         projection_log.append(f"❌ ERROR: Final projected FCFF (Year {forecast_yrs}) is invalid. TV calc failed."); projection_results['error'] = "Final FCFF invalid. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    else:
         if final_forecast_fcff <= 0:
             projection_log.append(f"ℹ️ FCFF Year {forecast_yrs} ({format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)}) <= 0. Setting TV=0.");
             terminal_value = 0.0; pv_terminal_value = 0.0
         else:
              fcff_n_plus_1 = final_forecast_fcff * (1 + terminal_growth_rate)
              terminal_value = fcff_n_plus_1 / (disc_rate - terminal_growth_rate)
              if not np.isfinite(terminal_value) or terminal_value < 0:
                  projection_log.append(f"⚠️ Calculated TV invalid/negative ({format_currency(terminal_value, currency_sym, is_indian=is_indian)}). Setting TV=0.")
                  terminal_value = 0.0; pv_terminal_value = 0.0
              else:
                   pv_factor_tv = 1 / ((1 + disc_rate) ** forecast_yrs); pv_terminal_value = terminal_value * pv_factor_tv
                   projection_log.append(f"FCFF Yr {forecast_yrs}: {format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)} -> FCFF Yr {forecast_yrs+1}: {format_currency(fcff_n_plus_1, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"Terminal Value (Yr {forecast_yrs}): {format_currency(terminal_value, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"PV Factor TV: {pv_factor_tv:.4f} -> PV(TV): {format_currency(pv_terminal_value, currency_sym, is_indian=is_indian)}")

    projection_results['terminal_value'] = terminal_value if np.isfinite(terminal_value) else 0.0
    projection_results['pv_terminal_value'] = pv_terminal_value if np.isfinite(pv_terminal_value) else 0.0

    display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
    return projection_results

# --- Execute FCFF calculation and projection ---
# Initialize fcff_results dictionary to ensure it's defined
fcff_results = {'error': 'Calculation not run yet.', 'initial_growth_rate': np.nan, 'growth_source':'N/A', 'terminal_growth_rate': np.nan}

if np.isfinite(discount_rate):
    try:
        # Pass currency symbol, flag, default growth rate, AND long history
        fcff_results = calculate_and_project_fcff(
            stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'], stock_data['history_long'], # Pass history_long
            wacc_components, discount_rate, FORECAST_YEARS, risk_free_rate_input, DEFAULT_GROWTH_RATE,
            currency_symbol, is_indian_stock)
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during FCFF Projection: {e}**"))
         fcff_results = {'error': f'Runtime error in FCFF calc: {e}', 'initial_growth_rate': np.nan, 'growth_source':'Error', 'terminal_growth_rate': np.nan}

else:
    display(Markdown("❌ **Skipping FCFF Projection due to invalid WACC.**"));
    fcff_results = {'error': 'Invalid WACC.', 'initial_growth_rate': np.nan, 'growth_source':'N/A (WACC Invalid)', 'terminal_growth_rate': np.nan}


# %% [markdown]
# ## 7. Enterprise Value and Equity Value Calculation

# %%
# Valuation Function Definition (No changes here)
def calculate_final_valuation(fcff_data, balance_sheet_df, shares_outstanding_val, currency_sym, wacc_data, is_indian): # Added flag
    display(Markdown("### Final Valuation Calculation..."))
    valuation = {'enterprise_value': np.nan, 'equity_value': np.nan, 'intrinsic_value_per_share': np.nan, 'error': None}
    valuation_log = []

    # Ensure fcff_data is a dictionary before accessing keys
    if not isinstance(fcff_data, dict):
        error_msg = f"FCFF Data is not a dictionary (type: {type(fcff_data)}). Cannot proceed."
        valuation['error'] = error_msg
        valuation_log.append(f"❌ ERROR: {error_msg}")
        display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
        return valuation

    if fcff_data.get('error'):
        valuation['error'] = f"FCFF Error: {fcff_data.get('error', 'Unknown')}"; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation

    pv_forecast_sum = fcff_data.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_data.get('pv_terminal_value', np.nan)
    if not np.isfinite(pv_forecast_sum) or not np.isfinite(pv_tv):
         valuation['error'] = f"Invalid PV components (PV FCFF: {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}, PV TV: {format_currency(pv_tv, currency_sym, is_indian=is_indian)})."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation # Pass flag

    enterprise_value = pv_forecast_sum + pv_tv; valuation['enterprise_value'] = enterprise_value
    # ----> Pass is_indian flag to format_currency calls <----
    valuation_log.append(f"PV(Forecast FCFFs): {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"PV(Terminal Value):   {format_currency(pv_tv, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"↳ Enterprise Value (EV): {format_currency(enterprise_value, currency_sym, is_indian=is_indian)}")

    # Get Debt & Cash from BS (use Debt from WACC for consistency)
    total_debt_latest = 0.0; cash_and_equivalents = 0.0; minority_interest = 0.0
    debt_source_info = wacc_data.get('debt_calc_method', '?'); total_debt_latest = wacc_data.get('debt_value_used', 0.0) # Get debt used in WACC

    if not balance_sheet_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        cash_keys = ["Cash And Cash Equivalents", "Cash Cash Equivalents And Short Term Investments", "Cash Financial", "Cash", "Cash And Short Term Investments"]
        cash_and_equivalents = safe_get(latest_bs, cash_keys, "BS", default=0)
        valuation_log.append(f"  + Cash & Equiv (BS): {format_currency(cash_and_equivalents, currency_sym, is_indian=is_indian)}")
        minority_interest = safe_get(latest_bs, ["Minority Interest", "Noncontrolling Interest"], "BS", default=0)
        if minority_interest != 0.0:
            valuation_log.append(f"  - Minority Interest: {format_currency(minority_interest, currency_sym, is_indian=is_indian)}")
    else:
        valuation_log.append("⚠️ Balance Sheet empty.")
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        valuation_log.append(f"  + Cash & Equiv: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")

    if not all(np.isfinite(v) for v in [enterprise_value, total_debt_latest, cash_and_equivalents, minority_interest]):
         valuation['error'] = "Invalid component for EV->Equity bridge."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    equity_value = enterprise_value - total_debt_latest + cash_and_equivalents - minority_interest; valuation['equity_value'] = equity_value
    valuation_log.append(f"↳ Equity Value: {format_currency(equity_value, currency_sym, is_indian=is_indian)}")

    # Calculate Intrinsic Value Per Share
    intrinsic_value_per_share = np.nan
    if not np.isfinite(equity_value): valuation['error'] = f"Equity value invalid ({equity_value})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    elif shares_outstanding_val <= 0: valuation['error'] = f"Shares outstanding invalid ({shares_outstanding_val})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    else:
        intrinsic_value_per_share = equity_value / shares_outstanding_val;
        valuation_log.append(f"  ÷ Shares Out: {shares_outstanding_val:,.0f}")
        valuation_log.append(f"▶️ Intrinsic Value/Share: {format_currency(intrinsic_value_per_share, currency_sym, is_indian=is_indian)}")
    valuation['intrinsic_value_per_share'] = intrinsic_value_per_share if np.isfinite(intrinsic_value_per_share) else np.nan

    display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
    return valuation

# --- Execute Final Valuation ---
# Initialize final_valuation to ensure it's defined
final_valuation = {'error': 'Calculation not run yet.', 'intrinsic_value_per_share': np.nan}

# Check if fcff_results is a dictionary and doesn't contain an error before proceeding
if isinstance(fcff_results, dict) and not fcff_results.get('error'):
    try:
        final_valuation = calculate_final_valuation(
            fcff_results, stock_data['balance_sheet'], shares_outstanding,
            currency_symbol, wacc_components, is_indian_stock
        )
    except Exception as e:
        display(Markdown(f"❌ **CRITICAL ERROR during Final Valuation: {e}**"))
        final_valuation = {'error': f'Runtime error in final valuation: {e}', 'intrinsic_value_per_share': np.nan}
elif isinstance(fcff_results, dict): # If fcff_results has an error
     display(Markdown("❌ **Skipping Final Valuation due to FCFF errors.**"));
     final_valuation = {'error': fcff_results.get('error', 'Upstream FCFF error.'), 'intrinsic_value_per_share': np.nan}
else: # If fcff_results is not even a dictionary (major upstream failure)
     display(Markdown("❌ **Skipping Final Valuation due to critical upstream errors.**"));
     final_valuation = {'error': 'FCFF results dictionary not created.', 'intrinsic_value_per_share': np.nan}


# %% [markdown]
# ## 8. Financial Ratio Analysis

# %%
# Ratio Function Definition (No changes here)
def analyze_ratios(stock_info_dict, financials_df, balance_sheet_df, cashflow_df,
                   enterprise_val, current_price_val, shares_outstanding_val,
                   currency_sym, user_benchmark_dict, # Only user inputs
                   benchmark_source_dict, # Dict tracking source per ratio
                   wacc_data, is_indian): # Pass WACC data for consistent debt and flag

    """ Calculates ratios, compares to user/fallback benchmarks, uses WACC-consistent debt. """
    display(Markdown("### Financial Ratio Analysis..."))
    ratios = {} # Calculated ratios {RatioName: value}
    ratio_log = []
    ratio_data_for_table = [] # For HTML table display
    ratio_plotly_data = {'Ratio': [], 'Company': [], 'Benchmark': [], 'Benchmark Source': [], 'Status': []} # For Plotly chart
    company_nm = stock_info_dict.get("shortName", ticker_symbol)

    # --- Pre-checks ---
    valid_price = current_price_val is not None and np.isfinite(current_price_val) and current_price_val > 0
    valid_shares = shares_outstanding_val is not None and np.isfinite(shares_outstanding_val) and shares_outstanding_val > 0
    # Ensure enterprise_val is valid before using it
    valid_enterprise_val = enterprise_val is not None and np.isfinite(enterprise_val)

    # Check for essential statements
    if financials_df.empty or balance_sheet_df.empty:
         ratio_log.append("❌ ERROR: Income Statement or Balance Sheet empty. Cannot calculate most ratios.")
         display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
         return {}, [] # Return empty results

    latest_fin = financials_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0] if not cashflow_df.empty else pd.Series(dtype=float)

    # --- Calculate Ratio Components (using safe_get with NaN default where appropriate) ---
    net_income = safe_get(latest_fin, ["Net Income Continuous Operations", "Net Income", "Net Income Common Stockholders", "Net Income Applicable To Common Shares"], default=np.nan)
    revenue = safe_get(latest_fin, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
    depreciation = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0) # Default 0 ok here
    ebitda = (ebit + depreciation) if np.isfinite(ebit) and np.isfinite(depreciation) else np.nan

    total_equity = safe_get(latest_bs, ["Total Stockholder Equity", "Stockholders Equity", "Total Equity Gross Minority Interest", "Common Stock Equity"], default=np.nan)
    total_assets = safe_get(latest_bs, ["Total Assets"], default=np.nan)
    current_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=np.nan)
    current_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=np.nan)
    inventory = safe_get(latest_bs, ["Inventory"], default=0) # Default 0 ok here

    # --- Use Debt value consistent with WACC calculation ---
    total_debt_ratio = wacc_data.get('debt_value_used', 0.0); debt_source_info = wacc_data.get('debt_calc_method', '?');
    ratio_log.append(f"Using Debt = {format_currency(total_debt_ratio, currency_sym, is_indian=is_indian)} (Source: WACC - {debt_source_info})")

    # Per Share Metrics
    eps = (net_income / shares_outstanding_val) if valid_shares and np.isfinite(net_income) and net_income != 0 else np.nan
    bvps = (total_equity / shares_outstanding_val) if valid_shares and np.isfinite(total_equity) and total_equity != 0 else np.nan

    # Capital Employed (ROCE denominator)
    capital_employed = np.nan; ce_source = "N/A"
    if np.isfinite(total_assets) and np.isfinite(current_liab):
        capital_employed_def1 = total_assets - current_liab
        if np.isfinite(capital_employed_def1) and capital_employed_def1 > 1.0:
            capital_employed = capital_employed_def1; ce_source = "A-CL"
    if not np.isfinite(capital_employed):
        if np.isfinite(total_equity) and np.isfinite(total_debt_ratio):
             capital_employed_def2 = total_equity + total_debt_ratio
             if np.isfinite(capital_employed_def2) and capital_employed_def2 > 1.0:
                 capital_employed = capital_employed_def2; ce_source = "E+D"
    if not np.isfinite(capital_employed):
         ce_source = "Invalid/Unavailable"; ratio_log.append("⚠️ Capital Employed could not be reliably calculated. ROCE affected.")

    # --- Calculate Ratios (Result is NaN if components invalid/zero denominator) ---
    ratios['P/E'] = (current_price_val / eps) if valid_price and np.isfinite(eps) and eps != 0 else np.nan
    ratios['P/B'] = (current_price_val / bvps) if valid_price and np.isfinite(bvps) and bvps != 0 else np.nan
    # Check valid_enterprise_val before EV/EBITDA
    ratios['EV/EBITDA'] = (enterprise_val / ebitda) if valid_enterprise_val and np.isfinite(ebitda) and ebitda != 0 else np.nan
    ratios['ROE'] = (net_income / total_equity) if np.isfinite(net_income) and np.isfinite(total_equity) and total_equity != 0 else np.nan
    ratios['ROA'] = (net_income / total_assets) if np.isfinite(net_income) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['ROCE'] = (ebit / capital_employed) if np.isfinite(ebit) and np.isfinite(capital_employed) and capital_employed != 0 else np.nan
    ratios['Asset Turnover'] = (revenue / total_assets) if np.isfinite(revenue) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['Current Ratio'] = (current_assets / current_liab) if np.isfinite(current_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan
    quick_assets = (current_assets - inventory) if np.isfinite(current_assets) and np.isfinite(inventory) else np.nan
    ratios['Quick Ratio'] = (quick_assets / current_liab) if np.isfinite(quick_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan

    # --- Compare with User/Fallback Benchmarks & Prepare Data Tables ---
    display(Markdown("*(Comparing ratios to user/fallback benchmarks...)*"))
    for key in REQUIRED_BENCHMARK_RATIOS:
        company_value = ratios.get(key); is_percent_type = key in ['ROE', 'ROA', 'ROCE']
        value_fmt = format_percent(company_value, 1) if is_percent_type else format_currency(company_value, '', 2, is_indian=False)
        if value_fmt == "N/A" and company_value is not None and not math.isfinite(company_value): value_fmt = "Inf"

        user_benchmark = user_benchmark_dict.get(key, None); fallback_benchmark = DEFAULT_BENCHMARK_RATIOS.get(key, None)
        benchmark_source = benchmark_source_dict.get(key, "?"); final_benchmark = None
        if benchmark_source == 'User Input' and user_benchmark is not None: final_benchmark = user_benchmark
        elif benchmark_source == 'Fallback Default' and fallback_benchmark is not None: final_benchmark = fallback_benchmark
        else: benchmark_source = "Not Available"

        bench_fmt = "N/A"
        if final_benchmark is not None: bench_fmt = f"{final_benchmark:.1f}%" if is_percent_type else f"{final_benchmark:.2f}"; bench_fmt += f" <i style='font-size:0.8em; color:grey'>({benchmark_source.replace('Fallback ','')})</i>"
        else: bench_fmt = "<i style='color:grey'>Not Provided</i>"

        status = "<i style='color:grey'>N/A</i>"; status_color = "grey"
        if np.isfinite(company_value) and final_benchmark is not None and np.isfinite(final_benchmark):
            numeric_value = company_value; numeric_benchmark = final_benchmark
            good_color = '#2ecc71'; bad_color = '#e74c3c'; neutral_color = '#f39c12'; grey_color='#7f8c8d'
            if abs(numeric_benchmark) > 1e-9: ratio_diff = (numeric_value - numeric_benchmark) / abs(numeric_benchmark)
            else: ratio_diff = float('inf') if numeric_value > 0 else float('-inf') if numeric_value < 0 else 0.0
            if key in ['ROE', 'ROA', 'ROCE', 'Asset Turnover', 'Current Ratio', 'Quick Ratio']: # Higher is better
                 if ratio_diff > 0.20: status_color = good_color; status = f"⬆️ Strong"
                 elif ratio_diff < -0.20: status_color = bad_color; status = f"⬇️ Weak"
                 else: status_color = grey_color; status = f"~ Average"
            elif key in ['P/E', 'P/B', 'EV/EBITDA']: # Lower might be better (valuation) or neutral
                 if ratio_diff < -0.20: status_color = good_color; status = f"✅ Low"
                 elif ratio_diff > 0.20: status_color = neutral_color; status = f"⚠️ High"
                 else: status_color = grey_color; status = f"~ Average"
            status = f"<span style='color:{status_color}'>{status}</span>"
            plot_value = numeric_value * 100 if is_percent_type else numeric_value; plot_benchmark = numeric_benchmark
            ratio_plotly_data['Ratio'].append(key); ratio_plotly_data['Company'].append(plot_value); ratio_plotly_data['Benchmark'].append(plot_benchmark);
            ratio_plotly_data['Benchmark Source'].append(benchmark_source); ratio_plotly_data['Status'].append(status)
        elif not np.isfinite(company_value): status = "<i style='color:grey'>Company N/A</i>"
        elif final_benchmark is None: status = "<i style='color:grey'>No Benchmark</i>"
        ratio_data_for_table.append([f"<b>{key}</b>", value_fmt, bench_fmt, status])

    # --- Display Logs & Table ---
    if ratio_log: display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
    if ratio_data_for_table:
        header = ["Ratio", f"{company_nm} Value", "Benchmark (Source)", "Status"]; html_table = "<p>(*Status vs User/Default benchmarks. Threshold: +/- 20%.)</p>"; html_table += "<table border='1' style='border-collapse:collapse; width:98%; margin-left:auto; margin-right:auto; font-size: 0.9em;'>"; html_table += "<thead style='background-color:#e9ecef; color: #495057;'><tr>" + "".join(f"<th style='padding: 8px; text-align:center;'>{h}</th>" for h in header) + "</tr></thead>"; html_table += "<tbody>"
        for i, row in enumerate(ratio_data_for_table): bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff"; html_table += f"<tr style='background-color:{bg_color};'>"; html_table += f"<td style='padding: 6px; text-align:left;'>{row[0]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[1]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[2]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[3]}</td>"; html_table += "</tr>"
        html_table += "</tbody></table><br>"; display(HTML(html_table))
    else: display(Markdown("*No ratio data available.*"))

    # --- Display Ratio Comparison Chart (Plotly Bullet) ---
    if ratio_plotly_data['Ratio']:
       num_ratios = len(ratio_plotly_data['Ratio']); fig_ratios = go.Figure()
       def get_range_colors(k): return ["#fadbd8", "#fdebd0", "#d5f5e3"] if k in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio'] else ["#d5f5e3","#fdebd0","#fadbd8"] if k in ['P/E','P/B','EV/EBITDA'] else ['lightgrey','darkgrey','grey']
       for i in range(num_ratios):
           ratio_name = ratio_plotly_data['Ratio'][i]; company_val = ratio_plotly_data['Company'][i]; benchmark_val = ratio_plotly_data['Benchmark'][i]; bench_src_plot = ratio_plotly_data['Benchmark Source'][i]
           if not np.isfinite(company_val) or not np.isfinite(benchmark_val): print(f"Skipping plot for {ratio_name}: Invalid Value ({company_val}) or Bench ({benchmark_val})."); continue
           range_colors = get_range_colors(ratio_name); is_percent = ratio_name in ['ROE', 'ROA', 'ROCE']
           poor_f=0.8; good_f=1.2; range1_limit = benchmark_val * (poor_f if benchmark_val >=0 else good_f); range2_limit = benchmark_val * (good_f if benchmark_val >=0 else poor_f)
           abs_max = max(abs(company_val), abs(benchmark_val)); plot_max_val = abs_max * 1.7 + 1; plot_min_val = 0
           if company_val < 0 or benchmark_val < 0: plot_min_val = min(company_val, benchmark_val) * 1.7 - 1
           if ratio_name in ['P/E', 'EV/EBITDA']: plot_max_val = min(plot_max_val, 100); plot_min_val = max(plot_min_val, -20)
           elif ratio_name in ['P/B']: plot_max_val = min(plot_max_val, 20); plot_min_val = max(plot_min_val, -5)
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: plot_max_val = min(plot_max_val, 5); plot_min_val = max(plot_min_val, 0)
           elif is_percent: plot_max_val = min(plot_max_val, 75); plot_min_val= max(plot_min_val,-50)
           else: plot_max_val = min(plot_max_val, 1000)
           if plot_min_val >= plot_max_val: plot_max_val = plot_min_val + 1
           steps = [];
           if abs(benchmark_val) < 1e-9: steps = [{'range': [plot_min_val, plot_max_val], 'color':'lightgrey'}]
           else: step_ranges = sorted([plot_min_val, range1_limit, range2_limit, plot_max_val]); step_colors = range_colors if benchmark_val >= 0 else range_colors[::-1]; steps = [{'range': [step_ranges[j], step_ranges[j+1]], 'color': step_colors[j]} for j in range(3)]
           bench_fmt_title = f"{benchmark_val:.1f}%" if is_percent else f"{benchmark_val:.2f}"; title_text = f"<b>{ratio_name}</b><br><i style='font-size:0.8em'>(Bench: {bench_fmt_title} [{bench_src_plot.replace('Fallback ','')}])</i>"
           lower_is_better = ratio_name in ['P/E', 'P/B', 'EV/EBITDA']; delta_decreasing_color = good_color if lower_is_better else bad_color; delta_increasing_color = bad_color if lower_is_better else good_color
           fig_ratios.add_trace(go.Indicator(
                mode = "number+gauge+delta", value = company_val, number={'suffix': "%" if is_percent else "", 'valueformat': ".1f"},
                delta = {'reference': benchmark_val, 'relative': True, 'valueformat':".0%", 'decreasing': {'color': delta_decreasing_color}, 'increasing': {'color': delta_increasing_color}},
                domain = {'row': i, 'column': 0}, title = {'text' : title_text, 'font':{'size':11}},
                gauge = { 'shape': "bullet", 'axis': {'range': [plot_min_val, plot_max_val] if plot_min_val < plot_max_val else [plot_min_val, plot_min_val+1], 'tickformat':'.1f'},
                    'threshold': { 'line': {'color': "black", 'width': 2}, 'thickness': 0.8, 'value': benchmark_val },
                    'steps': [ s for s in steps if s and isinstance(s.get('range'), list) and len(s['range'])==2 and s['range'][0] < s['range'][1] ],
                    'bar': {'color': "#2c3e50", 'thickness': 0.5} }
            ))
       fig_ratios.update_layout(height = max(300, num_ratios * 85), grid = {'rows': num_ratios, 'columns': 1, 'pattern': "independent"}, title_text='Ratio Performance vs. Benchmarks', title_x=0.5, margin=dict(l=160, r=40, t=60, b=30))
       fig_ratios.show()
    else: display(Markdown("*No valid ratios/benchmarks for comparison chart.*"))

    return ratios, ratio_data_for_table

# --- Execute Ratio Analysis ---
# Initialize ratios_calculated and ratio_table
ratios_calculated = {}; ratio_table = []

# Check if final_valuation is a dictionary and doesn't contain an error before proceeding
if isinstance(final_valuation, dict) and not final_valuation.get('error'):
    try:
        ratios_calculated, ratio_table = analyze_ratios(
            stock_data['info'], stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'],
            final_valuation.get('enterprise_value'), current_price, shares_outstanding, currency_symbol,
            user_benchmarks, benchmark_sources, wacc_components, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Ratio Analysis: {e}**"))
         ratios_calculated = {}; ratio_table = [] # Ensure they are reset on error
else:
    display(Markdown("❌ **Skipping Ratio Analysis due to prior errors.**"));
    ratios_calculated = {}; ratio_table = []


# %% [markdown]
# ## 9. Summary & Visualization Dashboard

# %%
# Dashboard Function Definition (No changes here, relies on valid inputs)
def display_summary_dashboard(company_nm, ticker_sym, current_pr, intrinsic_val_per_sh,
                              wacc_val, term_growth, init_growth, growth_src,
                              final_val_dict, hist_df, fcff_res, currency_sym, info_dict, input_src_dict,
                              is_indian): # Added flag

    display(Markdown(f"## Valuation Summary Dashboard: {company_nm} ({ticker_sym})"))
    source_summary = " | ".join(f"{k}:`{v.split('(')[0].strip()}`" for k, v in input_src_dict.items())
    # Display the source of the initial growth rate - Ensure growth_src is available
    growth_src_display = growth_src if growth_src else 'N/A'
    display(Markdown(f"**Input Sources:** {source_summary} | Growth:`{growth_src_display}`"))

    # Check validity of key inputs for display logic
    valid_intrinsic = intrinsic_val_per_sh is not None and np.isfinite(intrinsic_val_per_sh)
    valid_current_pr = current_pr is not None and np.isfinite(current_pr) and current_pr > 0
    valid_wacc = wacc_val is not None and np.isfinite(wacc_val)
    valid_term_g = term_growth is not None and np.isfinite(term_growth)
    valid_init_g = init_growth is not None and np.isfinite(init_growth)

    # --- Indicator Subplots ---
    fig_indicators = make_subplots(rows=1, cols=3, specs=[[{'type': 'indicator'}, {'type': 'indicator'}, {'type': 'indicator'}]],
                                 subplot_titles=("Price vs. Intrinsic", "WACC", "Growth Rates"), horizontal_spacing=0.1, column_widths=[0.4, 0.3, 0.3])

    # Indicator 1: Price vs Intrinsic
    intrinsic_val_num = intrinsic_val_per_sh if valid_intrinsic else np.nan
    intrinsic_val_fmt = format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian) if valid_intrinsic else "N/A"
    current_pr_num = current_pr if valid_current_pr else 0.0; upside_fmt = ""; delta_config = None
    if valid_intrinsic and valid_current_pr :
        if intrinsic_val_num > 0: upside_pct = (intrinsic_val_num / current_pr_num) - 1; upside_fmt = f" ({upside_pct:+.1%})"
        comparison_text = f"Intrinsic: {intrinsic_val_fmt}{upside_fmt}"
        delta_config = {'reference': intrinsic_val_num, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym, 'decreasing': {'color': '#2ecc71', 'symbol': '▲ Upside'}, 'increasing': {'color': '#e74c3c', 'symbol': '▼ Downside'}}
    elif valid_intrinsic: comparison_text = f"Intrinsic: {intrinsic_val_fmt} (Current N/A)"
    else: comparison_text = f"Intrinsic: N/A"
    fig_indicators.add_trace(go.Indicator(mode = "number+delta" if delta_config else "number", value = current_pr_num if valid_current_pr else 0, number={'prefix': currency_sym + " " if is_indian else currency_sym, 'valueformat': '.2f', 'font':{'size':24}}, title = {"text": f"<span style='font-size:0.8em'>{comparison_text}</span>"}, delta = delta_config, domain = {'row': 0, 'column': 0}), row=1, col=1)

    # Indicator 2: WACC Gauge
    wacc_val_num = wacc_val * 100 if valid_wacc else None
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = wacc_val_num,
        number = {'suffix': "%", 'valueformat':'.1f' if valid_wacc else '', 'font':{'size':20}},
        gauge = {'axis': {'range': [MIN_WACC*100, MAX_WACC*100]}, 'bar': {'color': "darkblue" if valid_wacc else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc", 'steps' : [ {'range': [MIN_WACC*100, 9], 'color': "#d5f5e3"}, {'range': [9, 14], 'color': "#fdebd0"}, {'range': [14, MAX_WACC*100], 'color': "#fadbd8"}]} if valid_wacc else None,
        title = {"text": "<span style='font-size:0.8em'>WACC</span>"}, domain = {'row': 0, 'column': 1}), row=1, col=2)

    # Indicator 3: Growth Gauge
    term_growth_num = term_growth * 100 if valid_term_g else None; init_growth_fmt = format_percent(init_growth, 1) if valid_init_g else "N/A"
    growth_title_text = f"<span style='font-size:0.8em'>Terminal Growth (g)<br>(Initial: {init_growth_fmt})</span>"
    gauge_max_range = max(TERMINAL_GROWTH_RATE_CAP * 100 + 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) + 1)
    gauge_min_range = MIN_GROWTH * 100 - 1
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = term_growth_num,
         number = {'suffix': "%", 'valueformat':'.1f' if valid_term_g else '', 'font':{'size':20}},
         gauge = {'axis': {'range': [gauge_min_range, gauge_max_range], 'tickwidth': 1}, 'bar': {'color': "darkgrey" if valid_term_g else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [{'range': [gauge_min_range, 2.0], 'color': "lightcoral"}, {'range': [2.0, 4.5], 'color': "lightyellow"}, {'range': [4.5, gauge_max_range], 'color': "lightgreen"}],
                 'threshold' : {'line': {'color': "black", 'width': 3}, 'thickness': 0.75, 'value': term_growth_num } if valid_term_g else None } if valid_term_g else None,
         title = {"text": growth_title_text}, domain = {'row': 0, 'column': 2}), row=1, col=3)

    fig_indicators.update_layout(height=220, margin=dict(l=30, r=30, t=50, b=20)); fig_indicators.show()

    # --- Historical Price Chart ---
    if not hist_df.empty:
        fig_hist = px.line(hist_df, x=hist_df.index, y='Close', title=f'{company_nm} Price History vs. DCF'); fig_hist.update_layout(xaxis_title='Date', yaxis_title=f'Price ({currency_sym})', height=400, legend_title_text='Legend'); fig_hist.update_xaxes(rangeslider_visible=False)
        if valid_intrinsic: fig_hist.add_hline(y=intrinsic_val_num, line_dash="dash", line_color="#e74c3c", opacity=0.8, annotation_text=f"DCF Intrinsic: {format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian)}", annotation_position="bottom right", annotation_font_color="#e74c3c")
        if valid_current_pr: fig_hist.add_hline(y=current_pr_num, line_dash="dot", line_color="#3498db", opacity=0.8, annotation_text=f"Current: {format_currency(current_pr_num, currency_sym, is_indian=is_indian)}", annotation_position="top left", annotation_font_color="#3498db")
        fig_hist.show()
    else: display(Markdown("*Historical price data unavailable.*"))

    # --- DCF EV Breakdown Pie Chart ---
    # Ensure fcff_res is a dictionary before accessing
    if isinstance(fcff_res, dict):
        pv_forecast_sum = fcff_res.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_res.get('pv_terminal_value', np.nan)
        valid_pvs = np.isfinite([pv_forecast_sum, pv_tv]); total_ev = (pv_forecast_sum + pv_tv) if all(valid_pvs) else np.nan
        if np.isfinite(total_ev) and total_ev != 0 :
            labels_pie = []; values_pie = []; pull_pie = []; colors = ['#4e79a7','#f28e2b']
            if abs(pv_forecast_sum) > 1e-6: labels_pie.append(f'PV({FORECAST_YEARS}yr FCFFs)'); values_pie.append(pv_forecast_sum); pull_pie.append(0)
            if abs(pv_tv) > 1e-6: labels_pie.append('PV(TV)'); values_pie.append(pv_tv); pull_pie.append(0.05 if len(labels_pie) > 1 else 0)
            if labels_pie:
                hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                if is_indian: hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                fig_dcf_pie = go.Figure(data=[go.Pie(labels=labels_pie, values=values_pie, hole=.4, pull=pull_pie, marker_colors=colors[:len(labels_pie)], hovertemplate=hover_format, textinfo='percent+label', insidetextorientation='radial', sort=False)])
                fig_dcf_pie.update_layout( title_text=f"DCF EV Breakdown (Total EV: {format_currency(total_ev, currency_sym, is_indian=is_indian)})", title_x=0.5, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), height=350, margin=dict(t=60, b=50, l=20, r=20) )
                fig_dcf_pie.show()
            else: display(Markdown(f"*DCF components zero/negligible. Cannot generate EV Breakdown chart.*"))
        else: display(Markdown(f"*EV calculation failed ({format_currency(total_ev, currency_sym, is_indian=is_indian)}). Cannot generate EV Breakdown chart.*"))
    else: display(Markdown(f"*FCFF results invalid. Cannot generate EV Breakdown chart.*"))

    # --- FCFF Projection Bar Chart ---
    # Ensure fcff_res is a dictionary before accessing
    if isinstance(fcff_res, dict) and fcff_res.get('projected_fcffs'):
        years = [f'Y{i+1}' for i in range(len(fcff_res['projected_fcffs']))]; pv_list_clean = [p if np.isfinite(p) else 0 for p in fcff_res.get('pv_fcffs', [])]; fcff_list_clean = [f if np.isfinite(f) else 0 for f in fcff_res.get('projected_fcffs', [])]
        if len(pv_list_clean) == len(years) and len(fcff_list_clean) == len(years):
            df_fcff_plot = pd.DataFrame({'Year': years, 'Forecasted FCFF': fcff_list_clean, 'PV of FCFF': pv_list_clean}); df_fcff_plot['Forecasted FCFF'] = pd.to_numeric(df_fcff_plot['Forecasted FCFF']); df_fcff_plot['PV of FCFF'] = pd.to_numeric(df_fcff_plot['PV of FCFF'])
            df_fcff_melt = df_fcff_plot.melt(id_vars='Year', var_name='Cash Flow Type', value_name='Value')
            fig_fcff_bar = px.bar(df_fcff_melt, x='Year', y='Value', color='Cash Flow Type', title='FCFF Forecast & Present Values', labels={'Value':f'Value ({currency_sym})'}, barmode='group', height=350, color_discrete_map={'Forecasted FCFF':'#5dade2', 'PV of FCFF':'#af7ac5'})
            hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            if is_indian: hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            fig_fcff_bar.update_traces(hovertemplate=hover_bar_format); fig_fcff_bar.update_layout(margin=dict(t=60, b=30, l=50, r=30)); fig_fcff_bar.show()
        else: display(Markdown("*FCFF projection data mismatch/invalid. Cannot plot.*"))
    else: display(Markdown("*FCFF projection data unavailable or invalid.*"))

    # --- Valuation Conclusion ---
    display(Markdown("### Valuation Conclusion"));
    # Ensure final_val_dict is a dictionary before accessing
    if isinstance(final_val_dict, dict):
        intrinsic_val = final_val_dict.get('intrinsic_value_per_share', np.nan)
        current_pr_num = current_pr if valid_current_pr else np.nan; conclusion = ""
        if final_val_dict.get('error'): conclusion = f"<span style='color:red'>❌ Valuation Error: {final_val_dict.get('error', '?')}</span>"
        elif not np.isfinite(intrinsic_val) or intrinsic_val <= 0: conclusion = f"⚠️ Valuation resulted in invalid/non-positive Intrinsic Value ({format_currency(intrinsic_val, currency_sym, is_indian=is_indian)})."
        elif not np.isfinite(current_pr_num): conclusion = f"✅ Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}. Cannot compare (Current Price N/A)."
        else:
             diff_pct = (intrinsic_val / current_pr_num) - 1
             current_fmt = format_currency(current_pr_num, currency_sym, is_indian=is_indian)
             intrinsic_fmt = format_currency(intrinsic_val, currency_sym, is_indian=is_indian)
             if abs(diff_pct) < 0.10: conclusion = f"<span style='color:grey'>~ Fairly Valued:</span> Price ({current_fmt}) vs DCF ({intrinsic_fmt}). Diff: {diff_pct:+.1%}"
             elif diff_pct < 0: overval_pct = (current_pr_num / intrinsic_val) - 1; conclusion = f"<span style='color:#e74c3c'>🔻 Potentially Overvalued:</span> Price ({current_fmt}) is **{overval_pct:+.1%}** above DCF ({intrinsic_fmt})."
             else: conclusion = f"<span style='color:#2ecc71'>✅ Potentially Undervalued:</span> Price ({current_fmt}) suggests **{diff_pct:+.1%}** upside to DCF ({intrinsic_fmt})."
    else:
        conclusion = f"<span style='color:red'>❌ Valuation Error: Final valuation data structure invalid.</span>"

    display(Markdown(f"> **{conclusion}**"))

# --- Execute Summary Dashboard Generation ---
# Check if results are valid dictionaries before passing them
if isinstance(final_valuation, dict) and isinstance(fcff_results, dict) and isinstance(stock_data.get('info'), dict) and isinstance(input_sources, dict) and not final_valuation.get('error'):
    try:
        display_summary_dashboard(
            company_name, ticker_symbol, current_price, final_valuation.get('intrinsic_value_per_share'),
            discount_rate, fcff_results.get('terminal_growth_rate'), fcff_results.get('initial_growth_rate'),
            fcff_results.get('growth_source', 'N/A'), final_valuation, stock_data['history'], fcff_results,
            currency_symbol, stock_data['info'], input_sources, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Dashboard Generation: {e}**"))
else:
    display(Markdown("❌ **Skipping Summary Dashboard due to prior errors or invalid data structures.**"))
    # Optionally print the error states for debugging:
    # print("Final Valuation State:", final_valuation)
    # print("FCFF Results State:", fcff_results)


# %% [markdown]
# ## 10. Assumptions & Disclaimer (Reflects Dynamic/User Inputs & Fallbacks - Growth Logic Updated)

# %%
# Dynamically generate assumption strings based on final values
# Add checks to ensure fcff_results and wacc_components are dictionaries before accessing
rf_source_disp = input_sources.get('Rf', '?')
rm_source_disp = input_sources.get('Rm','?')
beta_source_disp = input_sources.get('Beta','?')
c_spread_source_disp = input_sources.get('Credit Spread','N/A')
bench_source_disp = input_sources.get('Benchmarks','?')

growth_source_disp = fcff_results.get('growth_source', 'N/A') if isinstance(fcff_results, dict) else 'Error/Not Run'
tax_rate_source_disp = wacc_components.get('tax_rate_source', 'N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
equity_source_disp = wacc_components.get('equity_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
weights_source_disp = wacc_components.get('weights_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
terminal_growth_final = fcff_results.get('terminal_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan
initial_growth_final = fcff_results.get('initial_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan


# %%
# ----> UPDATED TEXT TO REFLECT NEW GROWTH LOGIC (V11) <----
display(Markdown(f"""
**Key Assumptions & Input Sources:**
*   **Financial Data:** Relies on data provided by Yahoo Finance (`yfinance`). Accuracy subject to provider. Missing data uses fallbacks or causes errors.
*   **Risk-Free Rate (Rf):** Attempted dynamic fetch, **confirmed/overridden by user.** (`Source: {rf_source_disp}` - Default suggestion: `{DEFAULT_RF_RATE_SUGGESTION:.1f}%`)
*   **Expected Market Return (Rm):** Based on **user input.** (`Source: {rm_source_disp}`)
*   **Market Risk Premium (MRP):** Derived from user `Rm` and final `Rf`.
*   **Beta (β):** Dynamically fetched if valid, otherwise **required user input.** (`Source: {beta_source_disp}`)
*   **Cost of Debt (Kd):** Implied if possible & within bounds (`{format_percent(MIN_COST_OF_DEBT)}` - `{format_percent(MAX_COST_OF_DEBT)}`). Otherwise, defaults to `Rf + Credit Spread`.
*   **Credit Spread:** Used only if default Kd triggered. Based on **user input** (default suggestion: `{format_percent(DEFAULT_CREDIT_SPREAD)}`). (`Source: {c_spread_source_disp}`)
*   **Growth Rates:**
    *   **Initial Growth:** `{format_percent(initial_growth_final)}` used for Year 1, determined via: (`Source: {growth_source_disp}`)
        1.  Fetch last 3 years (4 data points) of Revenue, FCFF, and Adjusted Stock Price.
        2.  **Validate Data:** Check if T-0 (end) is finite and T-3 (start) is positive and finite for each metric.
        3.  Calculate **3-Year CAGR** *only* for metrics with valid start/end data (using `calculate_cagr_modified`).
        4.  Calculate the **Geometric Mean** (`geometric_mean_modified`) of the successfully calculated CAGRs from step 3.
        5.  If GM calculation fails or *no valid CAGRs were computed* in step 3, use **Default (`{format_percent(DEFAULT_GROWTH_RATE)}`)**.
        6.  Apply **Bounds:** The result from step 4 or 5 is capped between **`{format_percent(MIN_INITIAL_GROWTH_FLOOR)}` (Floor)** and **`{format_percent(MAX_INITIAL_GROWTH_CEILING)}` (Ceiling)**.
    *   **Terminal Growth (g):** `{format_percent(terminal_growth_final)}` = min(GDP cap=`{format_percent(TERMINAL_GROWTH_RATE_CAP)}`, WACC*`0.67`, Final Initial Growth, Rf*`0.95`), floored at `{format_percent(MIN_GROWTH)}`, ensured `< WACC`.
*   **Tax Rate:** Effective rate `{tax_rate_source_disp}` (bounded `{format_percent(MIN_TAX_RATE)}` - `{format_percent(MAX_TAX_RATE)}`) or default (`{format_percent(DEFAULT_TAX_RATE)}`).
*   **Capital Structure:** Assumed stable using `{equity_source_disp}` and calculated debt. (`Weights Source: {weights_source_disp}`)
*   **Industry Benchmarks:** Uses **user input** or **generic fallbacks** if input left blank. (`Source: {bench_source_disp}`)
*   **Currency Formatting:** Uses ₹/Lakh/Crore for `.NS` tickers, standard K/M/B/T otherwise.

**Limitations:**
*   **Model Sensitivity:** DCF highly sensitive to inputs. **Sensitivity analysis (not included) crucial.**
*   **Input Quality:** Validity depends on user inputs and appropriateness of fallbacks.
*   **Historical Data Availability:** Financial statement data (needed for Rev/FCFF CAGR) is often limited to ~4 years via `yfinance`. Accuracy of 3-year CAGR depends heavily on these few points. Stock price history is usually more extensive. CAGR calculation is skipped if start/end data is invalid or start is non-positive.
*   **CAGR/GM Modifications:** The modified CAGR handles non-positive end values (relative to positive start) but requires a positive start value. The modified GM handles negative rates but falls back to arithmetic mean if rates are <= -100% or other errors occur. GM is less meaningful with highly disparate or negative rates.
*   **Dynamic Data Reliability:** Depends on `yfinance` accuracy/availability. Rf proxies need verification.
*   **Simplifications:** Linear decay, constant rates/structure. Ignores qualitative factors, macro shocks, sentiment.
*   **Fallback Accuracy:** Default growth rate (6.5%) and generic benchmarks may not be representative.
*   **Data Completeness:** Performance degrades with missing data. *Cross-verify key data.*

**Disclaimer:**
*For **educational and illustrative purposes only**. **NOT investment advice**. Models have limitations. Users MUST conduct own due diligence and consult a qualified financial advisor before investing. Creators NOT liable for actions based on this report.*
"""))


# %%



````

````python
# %% [markdown]
# # Dynamic Stock Financial Analysis & DCF Valuation
# **(Interactive Report - Dynamic/User Input Driven Assumptions - V13 - Robust Stock CAGR)**
#
# **Analyst:** Simulated AI Research Analyst
#
# **Objective:** To perform a comprehensive financial analysis and estimate the intrinsic value per share of a publicly traded company using the Discounted Cash Flow (DCF) methodology. Prioritizes dynamic fetching, falls back to user input, and uses configured defaults as a last resort. Includes fallback benchmarks for ratios and robust error handling/display. Adapts currency formatting for Indian stocks. **Initial growth uses a Geometric Mean of valid 3yr Rev/FCFF/Stock CAGRs (modified formulas for negatives), bounded & defaulted. Stock CAGR uses 'Adj Close' or 'Close' price.**
#
# **Methodology:**
# 1.  **User Input & Dynamic Fetching:** Get ticker. Attempt dynamic Rf, require user confirmation/input for Rf, Rm, Beta (if needed), Credit Spread (if needed), and Ratio Benchmarks (with fallbacks). Detects Indian stocks for currency formatting.
# 2.  **Data Acquisition:** Use `yfinance` for statements, info, market data. Fetch extended price history (4y, daily interval). **Inspect fetched columns.**
# 3.  **WACC Calculation:** Determine WACC using fetched/user inputs via CAPM. Handle errors.
# 4.  **FCFF Calculation & Projection:** Calculate historical FCFF. **Estimate initial growth by:**
#     *   Fetching last 3 years (4 data points) of Revenue, FCFF, and Stock Price (daily).
#     *   Checking validity of start (T-3) and end (T-0) points for each metric (start must be positive & finite, end finite).
#     *   Calculating 3-year CAGR *only* for metrics with valid data (using modified CAGR). **Stock CAGR now robustly uses 'Adj Close' if available, falling back to 'Close'.**
#     *   Calculating the Geometric Mean of the valid calculated CAGRs (using modified GM).
#     *   Falling back to default (6.5%) if GM fails or no valid CAGRs exist.
#     *   Bounding the result between 1.1% and 15%.
#     Project FCFF forward using linear decay to terminal growth. Handle errors.
# 5.  **Terminal Value:** Estimate value beyond the explicit forecast period.
# 6.  **Discounting & Valuation:** Discount cash flows and terminal value. Handle errors. Initialize result dicts.
# 7.  **Ratio Analysis:** Calculate ratios and compare to user-input or fallback default benchmarks. Handles non-finite values. Initialize result dicts.
# 8.  **Interactive Visualization:** Present findings using Plotly, including dynamic gauge ranges, robust display logic, and ₹/Lakh/Crore formatting for Indian stocks.

# %% [markdown]
# ## 1. Setup and Libraries

# %%
# Standard Libraries
import warnings
import sys
import numpy as np
import pandas as pd
import datetime
import math # For isnan, isinf checks

# Data Fetching
import yfinance as yf

# Visualization (Using Plotly)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

# Display Utilities
from IPython.display import display, Markdown, HTML

# Settings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pio.templates.default = "plotly_white" # Cleaner Plotly template

# %% [markdown]
# ## 2. Configuration and Helper Functions (Growth Helpers Updated)

# %%
# --- Model Configuration ---
DEFAULT_TICKER = "WIPRO.NS" # Default if user input fails
FORECAST_YEARS = 5
# Terminal Growth Rate CAP (Also acts as a default ceiling)
TERMINAL_GROWTH_RATE_CAP = 0.05 # Default/Cap set to 5%
# ----> DEFAULT INITIAL GROWTH (Used only if GM of CAGRs calc fails) <----
DEFAULT_GROWTH_RATE = 0.065  # Default set to 6.5%
# Default Risk-Free Rate suggestion (used if dynamic fetch fails)
DEFAULT_RF_RATE_SUGGESTION = 6.0 # Default set to 6.0% (as percent)

DEFAULT_CURRENCY_SYMBOL = "₹" # Changed default to Rupee, will be overridden by yfinance if possible
DEFAULT_TAX_RATE = 0.25     # General fallback tax rate if calculation fails
DEFAULT_CREDIT_SPREAD = 0.03 # Fallback spread over Rf for Cost of Debt
DEFAULT_MARKET_RISK_PREMIUM = 0.08 # Used only to suggest Rm to user

# --- Generic Fallback Benchmark Ratios (Used ONLY if User leaves input blank) ---
DEFAULT_BENCHMARK_RATIOS = {
    'P/E': 20.0,
    'P/B': 2.5,
    'EV/EBITDA': 12.0,
    'ROE': 15.0, # As percent value (e.g., 15 for 15%)
    'ROA': 8.0,  # As percent value
    'ROCE': 12.0,# As percent value
    'Asset Turnover': 0.8,
    'Current Ratio': 1.5,
    'Quick Ratio': 1.0
}

# --- Model Stability Bounds --- (Not primary financial assumptions, but guardrails)
MIN_WACC = 0.06
MAX_WACC = 0.25
MIN_TAX_RATE = 0.10
MAX_TAX_RATE = 0.40
MIN_GROWTH = 0.00 # Minimum growth floor (for terminal growth and projection steps)
# ----> Bounds for CALCULATED INITIAL GROWTH RATE <----
MIN_INITIAL_GROWTH_FLOOR = 0.011 # 1.1%
MAX_INITIAL_GROWTH_CEILING = 0.15  # 15.0%

MIN_COST_OF_DEBT = 0.03
MAX_COST_OF_DEBT = 0.15

# --- Ratio Keys Needed for Benchmarking ---
REQUIRED_BENCHMARK_RATIOS = list(DEFAULT_BENCHMARK_RATIOS.keys()) # Use keys from default dict

# --- Helper Functions ---
def print_keys(statement, name):
    """Prints available keys in a yfinance statement for debugging"""
    print(f"\n--- Available Keys in {name} ---")
    if isinstance(statement, pd.DataFrame):
        valid_indices = [idx for idx in statement.index if idx is not None]
        print(", ".join(map(str, valid_indices)))
    elif isinstance(statement, dict):
        print(", ".join(statement.keys()))
    print("------------------------------------\n")

def safe_get(data, key_list, statement_name="Statement", default=0.0):
    """
    Safely retrieves a numeric value from yfinance data (Series or Dict).
    Tries multiple keys, returns default if not found/invalid. Handles NaN/inf better.
    """
    if not isinstance(key_list, list):
        key_list = [key_list]

    value = None
    for key in key_list:
      current_value = None
      raw_val = None
      if isinstance(data, pd.Series):
          if key in data.index:
              try:
                  raw_val = data.loc[key]
                  # Check if raw_val is already numeric and finite before converting
                  if isinstance(raw_val, (int, float)) and np.isfinite(raw_val):
                      current_value = float(raw_val)
                  # Handle Series case (sometimes yfinance returns single value Series)
                  elif isinstance(raw_val, pd.Series):
                       if not raw_val.empty and pd.notna(raw_val.iloc[0]) and np.isfinite(raw_val.iloc[0]):
                            current_value = float(raw_val.iloc[0])
                  # Handle string conversion only if necessary and check finiteness
                  elif pd.notna(raw_val):
                      try_val = float(raw_val)
                      if np.isfinite(try_val):
                          current_value = try_val
              except (KeyError, TypeError, IndexError, ValueError):
                   pass # Ignore errors for this specific key/conversion
      elif isinstance(data, dict):
          raw_val = data.get(key)
          if raw_val is not None:
              try:
                  if isinstance(raw_val, str) and raw_val.lower() in ['n/a', 'none', '']:
                      pass # Treat as invalid
                  else:
                      temp_val = float(raw_val)
                      if np.isfinite(temp_val): # Check for NaN/inf after conversion
                           current_value = temp_val
              except (TypeError, ValueError):
                   pass

      if current_value is not None:
            value = current_value
            break # Found a valid value

    if value is None:
        value = default # Use the provided default if no valid value found
    return value

def format_currency(value, symbol="₹", precision=2, is_indian=False):
    """
    Formats number as currency string, handles None/NaN/inf.
    Uses Lakh/Crore for Indian stocks if is_indian=True.
    """
    if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
    if abs(value) > 1e18: return "Inf" # Handle extremely large numbers as Inf

    sign = "-" if value < 0 else ""
    abs_val = abs(value)

    if is_indian:
        # Use ₹ symbol regardless of fetched symbol for consistency
        symbol = "₹"
        if abs_val >= 1e7:  # 1 Crore (10 million)
            formatted_val = f"{abs_val / 1e7:.{precision}f} Cr"
        elif abs_val >= 1e5:  # 1 Lakh (100 thousand)
            formatted_val = f"{abs_val / 1e5:.{precision}f} Lakh"
        else:
            formatted_val = f"{abs_val:,.{precision}f}" # Add comma separators for smaller Rupee values
        return f"{sign}{symbol} {formatted_val}" # Add space after symbol
    else:
        # Standard international formatting (K, M, B, T)
        if abs_val >= 1e12: formatted_val = f"{abs_val / 1e12:.{precision}f}T"
        elif abs_val >= 1e9: formatted_val = f"{abs_val / 1e9:.{precision}f}B"
        elif abs_val >= 1e6: formatted_val = f"{abs_val / 1e6:.{precision}f}M"
        elif abs_val >= 1e3: formatted_val = f"{abs_val / 1e3:.{precision}f}K"
        else: formatted_val = f"{abs_val:.{precision}f}"
        return f"{sign}{symbol}{formatted_val}" # Standard symbol prefix

def format_percent(value, precision=1):
     """Formats number as percentage string, handles None/NaN/inf."""
     if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
     if abs(value) > 1e5: return "Inf" # Handle extreme % values

     return f"{value * 100:.{precision}f}%"

def get_validated_float_input(prompt_message, default_value=None, allow_blank=False, min_val=None, max_val=None):
    """Gets and validates user input for a float value, handles defaults, blanks, and bounds."""
    while True:
        prompt_with_default = prompt_message
        if default_value is not None:
            prompt_with_default += f" [Default: {default_value}]: "
        elif allow_blank:
             prompt_with_default += f" [Leave blank to skip/use fallback]: "
        else:
             prompt_with_default += ": "

        try:
            user_input_str = input(prompt_with_default).strip()

            if not user_input_str:
                if allow_blank: return None
                if default_value is not None:
                    try: return float(default_value)
                    except ValueError: raise ValueError(f"Internal Error: Invalid default value '{default_value}'.")
                raise ValueError("Input cannot be empty.")

            if user_input_str.endswith('%'): # Allow user to add % sign optionally
                user_input_str = user_input_str[:-1]

            value = float(user_input_str)

            if min_val is not None and value < min_val:
                 raise ValueError(f"Value must be >= {min_val}.")
            if max_val is not None and value > max_val:
                 raise ValueError(f"Value must be <= {max_val}.")

            return value
        except ValueError as e:
            print(f"❌ Invalid input: {e}. Please enter a valid number.")

def try_get_dynamic_rf(rf_ticker_symbol="^TNX"):
    """Attempts to fetch a dynamic risk-free rate (e.g., US 10Y Treasury)."""
    print(f"Attempting to fetch dynamic Risk-Free Rate ({rf_ticker_symbol})...")
    try:
        rf_ticker = yf.Ticker(rf_ticker_symbol)
        hist = rf_ticker.history(period="5d")
        if not hist.empty and 'Close' in hist.columns:
            last_close = hist['Close'].iloc[-1]
            if pd.notna(last_close):
                dynamic_rf = last_close / 100.0 # Assume Yahoo gives yield as %
                if 0.001 < dynamic_rf < 0.20: # Check bounds 0.1% to 20%
                    print(f"✅ Dynamically fetched Rf ({rf_ticker_symbol}): {format_percent(dynamic_rf)}")
                    return dynamic_rf
                else:
                     print(f"⚠️ Fetched Rf value ({format_percent(dynamic_rf)}) out of typical range. Ignoring.")
                     return None
            else:
                 print(f"⚠️ Fetched Rf data was invalid (NaN) for {rf_ticker_symbol}.")
                 return None
        else:
             print(f"⚠️ Could not fetch recent data for {rf_ticker_symbol}.")
             return None
    except Exception as e:
        print(f"❌ Error fetching dynamic Rf ({rf_ticker_symbol}): {e}")
        return None

# --- Growth Calculation Helpers (Updated) ---
# calculate_average_growth kept for reference
def calculate_average_growth(values_list):
    """Calculates average simple growth rate (NOT CAGR). Kept for reference."""
    growth_rates = []
    if len(values_list) < 2: return np.nan
    for i in range(len(values_list) - 1):
        val_t = values_list[i]; val_t_minus_1 = values_list[i+1]
        if pd.notna(val_t) and pd.notna(val_t_minus_1) and math.isfinite(val_t) and math.isfinite(val_t_minus_1):
            if abs(val_t_minus_1) > 1e-9:
                g = (val_t / val_t_minus_1) - 1
                if -10.0 < g < 10.0: growth_rates.append(g)
    return np.mean(growth_rates) if growth_rates else np.nan

def calculate_cagr_modified(start_val, end_val, num_years):
    """
    Calculates Compound Annual Growth Rate (CAGR).
    MODIFIED to handle cases where end_val is negative or zero, assuming start_val is positive.
    Returns NaN if start_val is non-positive, num_years <= 0, or values are invalid.
    Also returns NaN if result is extremely high/low (often indicates data issues).
    """
    if pd.isna(start_val) or pd.isna(end_val) or not math.isfinite(start_val) or not math.isfinite(end_val) or num_years <= 0:
        return np.nan
    # CRITICAL: Start value must be positive for any meaningful CAGR interpretation.
    if start_val <= 1e-9:
        return np.nan

    cagr = np.nan # Initialize cagr as nan
    try:
        # Handle non-positive end value
        if end_val <= 1e-9:
            end_val_abs = max(abs(end_val), 1e-9)
            ratio = end_val_abs / start_val
            if ratio > 0:
                 # Calculate magnitude and make it negative
                 cagr_magnitude = ratio ** (1 / num_years) - 1
                 cagr = -cagr_magnitude
            # else: cagr remains NaN
        else:
            # Standard CAGR calculation if both start and end are positive
            ratio = end_val / start_val
            if ratio > 0:
                 cagr = ratio ** (1 / num_years) - 1
            # else: cagr remains NaN

    except (ValueError, OverflowError, ZeroDivisionError) as e:
        cagr = np.nan # Ensure cagr is NaN on error

    # Final check for extremely large/small results
    if pd.notna(cagr) and not (-0.9999 < cagr < 10.0): # Reject >1000% or <-99.99% growth
        cagr = np.nan

    return cagr


def geometric_mean_modified(rates):
    """
    Calculates the geometric mean of a list of rates (e.g., CAGRs).
    MODIFIED using an offset method to handle negative rates (down to nearly -100%).
    Falls back to arithmetic mean if any rate is <= -1.0 (-100%) or calculation fails.
    Returns NaN if no valid rates are provided.
    """
    # Filter out non-finite values first
    valid_rates = [r for r in rates if pd.notna(r) and math.isfinite(r)]
    if not valid_rates:
        return np.nan
    if len(valid_rates) == 1:
        return valid_rates[0] # GM of one item is the item itself

    # Offset method: Add 1 to each rate. This factor MUST be > 0.
    offset = 1.0
    factors_offset = [r + offset for r in valid_rates]

    # Check if any rate was <= -1.0 (which makes the factor <= 0)
    if any(f <= 1e-9 for f in factors_offset): # Use a small threshold
         print("⚠️ Geometric mean calculation failed due to rate <= -100%. Falling back to arithmetic mean.")
         return np.mean(valid_rates) # Fallback uses only the valid rates

    n = len(valid_rates)
    product = 1.0
    try:
        for factor in factors_offset:
             product *= factor
        # Handle potential complex numbers if product is negative and n is even
        if product < 0 and n % 2 == 0:
             print("⚠️ Geometric mean calculation failed (negative base for even root). Falling back to arithmetic mean.")
             return np.mean(valid_rates)

        # Calculate GM of offset rates (sign handling for odd roots of negative numbers)
        gm_offset = np.sign(product) * (np.abs(product) ** (1/n))

        # Subtract the offset
        gm = gm_offset - offset

        # Add a check for extreme GM results
        if pd.notna(gm) and not (-0.9999 < gm < 10.0):
            print(f"⚠️ Geometric mean result ({format_percent(gm)}) out of reasonable bounds. Falling back to arithmetic mean.")
            return np.mean(valid_rates)

        return gm
    except (ValueError, OverflowError, ZeroDivisionError, TypeError) as e: # Added TypeError
        print(f"⚠️ Geometric mean calculation encountered numeric error ({e}). Falling back to arithmetic mean.")
        return np.mean(valid_rates)

# --- End Growth Calculation Helpers ---


# %% [markdown]
# ## 3. User Input & Dynamic Fetching: Ticker, Rf, Rm, Benchmarks, Beta, Spread

# %%
input_sources = {}

# --- Get Ticker Symbol ---
user_ticker = input(f"Enter stock ticker (e.g., MSFT, AAPL, RELIANCE.NS) [Default: {DEFAULT_TICKER}]: ")
if not user_ticker: user_ticker = DEFAULT_TICKER
ticker_symbol = user_ticker.strip().upper()
display(Markdown(f"**Analyzing Ticker:** `{ticker_symbol}`"))

# ----> DETECT IF INDIAN STOCK for currency formatting <----
is_indian_stock = ticker_symbol.endswith(".NS")
if is_indian_stock:
    print("Info: Detected Indian stock (.NS). Using ₹/Lakh/Crore formatting.")

# --- Determine RF Ticker based on Stock Ticker ---
rf_proxy_ticker = "^TNX" # Default US 10Y
if is_indian_stock: rf_proxy_ticker = "^NS10"; print(f"Info: Using Indian Rf proxy: {rf_proxy_ticker}")
elif ticker_symbol.endswith(".L"): rf_proxy_ticker = "^FTLC"; print(f"Info: Using UK Rf proxy: {rf_proxy_ticker}")
# Add more country/exchange specific Rf proxies if needed

# --- Get Risk-Free Rate (Rf) ---
print("\n--- CAPM Inputs ---")
dynamic_rf_rate = try_get_dynamic_rf(rf_proxy_ticker)
rf_prompt = f"Enter Risk-Free Rate (Rf) as percentage (e.g., '4.5' for 4.5%). Fetched {rf_proxy_ticker}?"
suggested_rf_val = dynamic_rf_rate * 100 if dynamic_rf_rate is not None else DEFAULT_RF_RATE_SUGGESTION

risk_free_rate_input_pct = get_validated_float_input(
    rf_prompt, default_value=f"{suggested_rf_val:.2f}", allow_blank=False, min_val=0.0, max_val=20.0
)
risk_free_rate_input = risk_free_rate_input_pct / 100.0

input_sources['Rf'] = 'Dynamic Fetch (User Confirmed/Adjusted)' if dynamic_rf_rate is not None else f'User Input (Default: {DEFAULT_RF_RATE_SUGGESTION:.1f}%)'
display(Markdown(f"**Using Risk-Free Rate (Rf):** {format_percent(risk_free_rate_input)} (`Source: {input_sources['Rf']}`)"))

# --- Get Expected Market Return (Rm) ---
suggested_mrp = DEFAULT_MARKET_RISK_PREMIUM
suggested_rm = risk_free_rate_input + suggested_mrp
rm_prompt = f"Enter Expected Market Return (Rm) as percentage (e.g., '10'). Suggestion based on Rf + {format_percent(suggested_mrp)} MRP is {format_percent(suggested_rm)}"

while True:
    market_return_input_pct = get_validated_float_input(
        rm_prompt, default_value=f"{suggested_rm * 100:.1f}", allow_blank=False, min_val=-10.0, max_val=35.0
    )
    market_return_input = market_return_input_pct / 100.0
    if market_return_input >= risk_free_rate_input: break
    else: print(f"❌ Error: Rm ({format_percent(market_return_input)}) must be >= Rf ({format_percent(risk_free_rate_input)}). Re-enter.")

market_risk_premium_input = market_return_input - risk_free_rate_input
input_sources['Rm'] = 'User Input'
input_sources['MRP'] = 'Derived'
display(Markdown(f"**Using Expected Market Return (Rm):** {format_percent(market_return_input)} (`Source: {input_sources['Rm']}`)"))
display(Markdown(f"**↳ Implied Market Risk Premium (MRP):** {format_percent(market_risk_premium_input)} (`Source: {input_sources['MRP']}`)"))

# --- Beta (Placeholder) ---
beta_input = None; input_sources['Beta'] = 'Pending'
# --- Credit Spread (Placeholder) ---
user_credit_spread = None; input_sources['Credit Spread'] = 'Pending'

# --- Get Industry Benchmark Ratios (with fallback defaults) ---
print("\n--- Industry Ratio Benchmarks ---")
display(Markdown("Enter benchmarks or leave blank for generic fallbacks. *(Find industry averages online)*"))
user_benchmarks = {}; benchmark_sources = {}
print("Provide benchmarks (e.g., ROE: enter '15' for 15%; P/E: enter '20'):")
for ratio_key in REQUIRED_BENCHMARK_RATIOS:
    default_val = DEFAULT_BENCHMARK_RATIOS.get(ratio_key, None)
    prompt = f"  Enter benchmark for {ratio_key}"
    is_percent_type = ratio_key in ['ROE', 'ROA', 'ROCE']
    min_b, max_b = (-50, 100) if is_percent_type else (-100, 500)

    bench_val_input = get_validated_float_input(
        prompt, default_value=f"{default_val:.2f}" if default_val is not None else None,
        allow_blank=True, min_val=min_b, max_val=max_b
    )

    if bench_val_input is not None:
        user_benchmarks[ratio_key] = bench_val_input
        benchmark_sources[ratio_key] = 'User Input'
        display_val = f"{bench_val_input:.1f}%" if is_percent_type else f"{bench_val_input:.2f}"
        display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: User`)"))
    elif default_val is not None:
         user_benchmarks[ratio_key] = default_val # Store the default value if user skipped
         benchmark_sources[ratio_key] = 'Fallback Default'
         display_val = f"{default_val:.1f}%" if is_percent_type else f"{default_val:.2f}"
         display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: Default`)"))
    else:
        benchmark_sources[ratio_key] = 'Not Available'
        display(Markdown(f"  - **{ratio_key} Benchmark:** Not Provided / No Default"))
input_sources['Benchmarks'] = 'User Input or Default'

# %% [markdown]
# ## 4. Data Acquisition (Fetching Extended History)

# %%
stock_data = {}
try:
    stock = yf.Ticker(ticker_symbol)
    info = stock.info
    if not info or not info.get('regularMarketPrice'):
        print(f"Warning: Limited info for {ticker_symbol}. Trying history...")
        hist_check = stock.history(period="5d");
        if hist_check.empty: raise ValueError(f"Ticker '{ticker_symbol}' invalid/no data.")
        else: print(f"Warning: Proceeding with limited data."); info = {} if not isinstance(info, dict) else info
    stock_data['info'] = info

    # Standard statement fetching (usually 4 years)
    stock_data['financials'] = stock.financials
    stock_data['balance_sheet'] = stock.balance_sheet
    stock_data['cashflow'] = stock.cashflow
    if stock_data['financials'].empty or stock_data['balance_sheet'].empty or stock_data['cashflow'].empty:
       print(f"⚠️ Warning: Empty statements for {ticker_symbol}. DCF unreliable.")

    # Fetch recent history for charts/current price fallback
    stock_data['history'] = stock.history(period="2y")
    # ----> Fetch extended history (need at least ~3 years for 3yr CAGR) <----
    # Fetch 4 years daily data to maximize chance of finding start date ~3 years ago
    stock_data['history_long'] = stock.history(period="4y", interval="1d") # Explicitly daily
    # ----> INSPECT COLUMNS <----
    if not stock_data['history_long'].empty:
        print(f"Info: Fetched long history (4y daily). Columns available: {list(stock_data['history_long'].columns)}")
    else:
         print(f"⚠️ Warning: Could not fetch long history (4y daily) for {ticker_symbol}. Stock CAGR calc may fail.")
    if stock_data['history_long'].empty or len(stock_data['history_long']) < 2: # Need at least 2 points
        print(f"⚠️ Warning: Fetched long history has < 2 data points. Stock CAGR calc will fail.")


    company_name = info.get("longName", info.get("shortName", ticker_symbol))
    # Determine currency symbol, prioritizing Rupee for Indian stocks
    currency_code = info.get("currency", DEFAULT_CURRENCY_SYMBOL)
    if not isinstance(currency_code, str) or len(currency_code)>5 : currency_code=DEFAULT_CURRENCY_SYMBOL
    currency_symbol = "₹" if is_indian_stock else info.get("symbol", currency_code) # Prefer symbol if available, else code, force Rupee for Indian

    current_price = safe_get(info, ['currentPrice', 'regularMarketPrice', 'previousClose', 'lastPrice'], 'Info', default=np.nan)
    price_source = "yfinance Info"
    if not np.isfinite(current_price) and not stock_data['history'].empty:
        last_close_hist = stock_data['history']['Close'].iloc[-1]
        if pd.notna(last_close_hist): current_price = last_close_hist; price_source = "yfinance History"
    if not np.isfinite(current_price): print(f"⚠️ Warning: Could not determine current price."); current_price = np.nan

    shares_outstanding = safe_get(info, ['sharesOutstanding'], 'Info', default=0)

    beta_input = safe_get(info, ['beta'], 'Info', default=None)
    if beta_input is not None and np.isfinite(beta_input):
        input_sources['Beta'] = 'Dynamic Fetch'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))
    else:
         print(f"❌ ERROR: Beta not found dynamically or invalid.")
         beta_input = get_validated_float_input(f"Enter Beta for {ticker_symbol}:", allow_blank=False, min_val=-1.0, max_val=4.0)
         info['beta'] = beta_input; input_sources['Beta'] = 'User Input'
         display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))

    market_cap = safe_get(info, ['marketCap'], 'Info', default=0)
    mcap_source = "yfinance Info"
    if market_cap <= 0 and shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Calculated"; print(f"Info: Market Cap calculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}") # Pass flag
    elif market_cap <= 0: mcap_source = "Unavailable"; print(f"⚠️ Warning: Market Cap unavailable."); market_cap = np.nan

    sector = info.get("sector", "N/A"); industry = info.get("industry", "N/A")
    summary = info.get("longBusinessSummary", "No summary available.")

    display(Markdown(f"**Company:** {company_name}")); display(Markdown(f"**Sector:** {sector} | **Industry:** {industry}"))
    display(Markdown(f"**Currency:** {currency_code} ({currency_symbol})")); # Show both code and effective symbol
    display(Markdown(f"**Current Market Price:** {format_currency(current_price, currency_symbol, is_indian=is_indian_stock)} (`Source: {price_source}`)"))
    display(Markdown(f"**Market Cap:** {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)} (`Source: {mcap_source}`)"))

except Exception as e: print(f"❌ CRITICAL ERROR fetching data: {e}"); sys.exit("Cannot proceed.")

if not isinstance(shares_outstanding, (int, float)) or shares_outstanding <= 0: sys.exit("Essential data missing: Shares Outstanding.")
if beta_input is None or not np.isfinite(beta_input): sys.exit("Essential data missing: Beta.")
if not np.isfinite(market_cap): print(f"⚠️ WARNING: Market Cap unavailable. WACC weights less accurate.")

# %% [markdown]
# ## 5. WACC Calculation

# %%
# WACC Function Definition (No changes here)
def calculate_wacc(stock_info_dict, financials_df, balance_sheet_df,
                   user_rf, user_mrp, user_beta, input_sources_dict,
                   currency_sym, is_indian): # Added currency params
    global user_credit_spread
    display(Markdown("### Calculating WACC..."))
    components = {}
    calculation_log = []
    calculation_log.append(f"Input Sources: Rf='{input_sources_dict.get('Rf')}', MRP='{input_sources_dict.get('MRP')}', Beta='{input_sources_dict.get('Beta')}'")

    beta_val = user_beta
    if beta_val <= 0: calculation_log.append(f"⚠️ Warning: Non-positive Beta ({beta_val:.2f}).")
    components['beta'] = beta_val; components['risk_free_rate'] = user_rf; components['market_risk_premium'] = user_mrp
    cost_of_equity = user_rf + beta_val * user_mrp
    components['cost_of_equity'] = cost_of_equity
    calculation_log.append(f"Inputs -> Rf: {format_percent(user_rf)}, Beta: {beta_val:.2f}, MRP: {format_percent(user_mrp)}")
    calculation_log.append(f"↳ Cost of Equity (Ke) [CAPM]: {format_percent(cost_of_equity)}")

    cost_of_debt_pre_tax = np.nan; tax_rate = np.nan; after_tax_cost_of_debt = np.nan
    total_debt = 0.0; debt_calculation_method = "N/A"; tax_rate_source = "N/A"

    if not balance_sheet_df.empty and not financials_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]; latest_fin = financials_df.iloc[:, 0]
        debt_keys = ['Total Debt', 'Long Term Debt', 'Long Term Debt And Capital Lease Obligation', 'Total Debt Non Current']
        long_term_debt = safe_get(latest_bs, debt_keys , 'Balance Sheet', default=0)
        current_debt_keys = ['Current Debt', 'Short Term Debt', 'Current Debt And Capital Lease Obligation', 'Short Long Term Debt', 'Total Debt Current']
        short_term_debt = safe_get(latest_bs, current_debt_keys , 'Balance Sheet', default=0)
        total_debt = long_term_debt + short_term_debt
        components['debt_value_used'] = total_debt
        interest_expense = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], 'Income Statement', default=0))

        implied_rate_calculated = False
        if total_debt > 1000 and interest_expense > 0:
             implied_rate = interest_expense / total_debt; implied_rate_calculated = True
             if MIN_COST_OF_DEBT <= implied_rate <= MAX_COST_OF_DEBT:
                  cost_of_debt_pre_tax = implied_rate; debt_calculation_method = "Implied";
                  calculation_log.append(f"Interest Exp: {format_currency(interest_expense, symbol=currency_sym, is_indian=is_indian)}, Debt: {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
                  calculation_log.append(f"↳ Implied Cost of Debt (Kd pre-tax): {format_percent(cost_of_debt_pre_tax)}")
             else: calculation_log.append(f"⚠️ Implied Kd ({format_percent(implied_rate)}) out of bounds. Falling back.")
        elif total_debt > 1000: calculation_log.append(f"ℹ️ Found Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}) but no Interest Exp. Falling back.")

        if not debt_calculation_method.startswith("Implied"):
             if total_debt > 1000:
                 debt_calculation_method = "Default (Rf+Spread)"
                 if user_credit_spread is None:
                     spread_prompt = f"Implied Kd failed/NA. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
                     user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
                     user_credit_spread = user_credit_spread_pct / 100.0
                     input_sources_dict['Credit Spread'] = 'User Input (Default Kd)'
                 else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
                 cost_of_debt_pre_tax = user_rf + user_credit_spread
                 calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(cost_of_debt_pre_tax)}")
                 calculation_log.append(f"   (Spread Source: {input_sources_dict['Credit Spread']})")
             else:
                 debt_calculation_method = "Zero Debt"; cost_of_debt_pre_tax = 0.0; input_sources_dict['Credit Spread'] = 'N/A (Zero Debt)';
                 calculation_log.append(f"ℹ️ Negligible Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}). Kd = 0%.")

        income_tax_expense = safe_get(latest_fin, ['Tax Provision', 'Income Tax Expense'], 'Income Statement', default=0)
        pretax_income = safe_get(latest_fin, ['Pretax Income', 'EBT', 'Earnings Before Tax'], 'Income Statement', default=0)
        if pretax_income <= 0 and income_tax_expense > 0:
             ebit_val = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=0); interest_exp_for_ebt = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], default=0))
             pretax_income_from_ebit = ebit_val - interest_exp_for_ebt
             if pretax_income_from_ebit > 0: pretax_income = pretax_income_from_ebit
        if pretax_income > 0 and income_tax_expense >= 0:
            calculated_tax_rate = income_tax_expense / pretax_income; tax_rate = min(max(calculated_tax_rate, MIN_TAX_RATE), MAX_TAX_RATE); tax_rate_source = "Derived & Bounded"
            if abs(tax_rate - calculated_tax_rate) > 0.001: tax_rate_source = f"Derived ({format_percent(calculated_tax_rate)}), Bounded"
            calculation_log.append(f"Tax Exp: {format_currency(income_tax_expense, symbol=currency_sym, is_indian=is_indian)}, Pretax Inc: {format_currency(pretax_income, symbol=currency_sym, is_indian=is_indian)}")
            calculation_log.append(f"↳ Effective Tax Rate: {format_percent(tax_rate)} ({tax_rate_source})")
        else: tax_rate = DEFAULT_TAX_RATE; tax_rate_source = "Default"; calculation_log.append(f"⚠️ Cannot derive tax rate. Using default: {format_percent(tax_rate)}")
    else: # Financial statements empty
         calculation_log.append("⚠️ Financial statements empty. Using defaults."); debt_calculation_method = "Default (No Fin.)"; tax_rate_source = "Default (No Fin.)"
         if user_credit_spread is None:
              spread_prompt = f"No financials. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
              user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
              user_credit_spread = user_credit_spread_pct / 100.0; input_sources_dict['Credit Spread'] = 'User Input (No Fin.)'
         else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
         cost_of_debt_pre_tax = user_rf + user_credit_spread; tax_rate = DEFAULT_TAX_RATE; total_debt = 0; components['debt_value_used'] = 0
         calculation_log.append(f"Using Default Kd: {format_percent(cost_of_debt_pre_tax)} (Rf + Spread: {input_sources_dict['Credit Spread']})")
         calculation_log.append(f"Using Default Tax Rate: {format_percent(tax_rate)}")

    components['cost_of_debt'] = cost_of_debt_pre_tax if np.isfinite(cost_of_debt_pre_tax) else 0.0
    components['debt_calc_method'] = debt_calculation_method
    components['tax_rate'] = tax_rate if np.isfinite(tax_rate) else DEFAULT_TAX_RATE
    components['tax_rate_source'] = tax_rate_source
    if np.isfinite(cost_of_debt_pre_tax) and np.isfinite(tax_rate): after_tax_cost_of_debt = cost_of_debt_pre_tax * (1 - tax_rate)
    else: after_tax_cost_of_debt = components['cost_of_debt'] * (1 - DEFAULT_TAX_RATE) # Use default if calculation failed
    components['after_tax_cost_of_debt'] = after_tax_cost_of_debt
    if components['cost_of_debt'] > 0 : calculation_log.append(f"↳ Cost of Debt (After-tax): {format_percent(after_tax_cost_of_debt)}")

    market_cap_val = safe_get(stock_info_dict, ['marketCap'], 'Info', default=np.nan); equity_source = "Market Cap" if np.isfinite(market_cap_val) and market_cap_val > 0 else "N/A"
    equity_value_for_wacc = np.nan
    if np.isfinite(market_cap_val) and market_cap_val > 0: equity_value_for_wacc = market_cap_val
    else:
       calculation_log.append("⚠️ Market Cap unavailable/zero."); equity_source = "Book Equity Fallback"
       if not balance_sheet_df.empty:
           equity_keys = ['Total Stockholder Equity', 'Stockholders Equity', 'Total Equity Gross Minority Interest', 'Common Stock Equity']
           book_equity_val = safe_get(balance_sheet_df.iloc[:, 0], equity_keys, 'Balance Sheet', default=np.nan)
           if np.isfinite(book_equity_val) and book_equity_val > 1.0:
              equity_value_for_wacc = book_equity_val; calculation_log.append(f"Using Book Equity ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}) as proxy.") # Pass flag
           else: equity_source = "Placeholder (Book Invalid)"
       else: equity_source = "Placeholder (No BS)"
       if not np.isfinite(equity_value_for_wacc): equity_value_for_wacc = 1.0; calculation_log.append("⚠️ Market Cap & Book Equity unavailable. Using placeholder Equity Value (1.0). Weights inaccurate.")
    components['equity_value_used'] = equity_value_for_wacc; components['equity_source'] = equity_source

    equity_weight = np.nan; debt_weight = np.nan; weights_source = "N/A"; total_capital = equity_value_for_wacc + total_debt
    if np.isfinite(total_capital) and total_capital > 1.0:
        equity_weight = equity_value_for_wacc / total_capital; debt_weight = total_debt / total_capital; weights_source = "Calculated"
        total_weight = equity_weight + debt_weight # Normalize if calculation slightly off
        if abs(total_weight - 1.0) > 1e-6: equity_weight /= total_weight; debt_weight /= total_weight
    else: calculation_log.append(f"❌ ERROR: Total Capital ({format_currency(total_capital, symbol=currency_sym, is_indian=is_indian)}) invalid. Using default weights (70/30)."); equity_weight = 0.7; debt_weight = 0.3; weights_source = "Default" # Pass flag
    components['equity_weight'] = equity_weight; components['debt_weight'] = debt_weight; components['weights_source'] = weights_source
    calculation_log.append(f"Equity Value Used ({components['equity_source']}): {format_currency(components['equity_value_used'], symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"Debt Value Used ({components['debt_calc_method']}): {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"↳ Weights E/V: {format_percent(equity_weight)}, D/V: {format_percent(debt_weight)} (Source: {weights_source})")

    wacc = np.nan
    if all(np.isfinite(v) for v in [equity_weight, cost_of_equity, debt_weight, after_tax_cost_of_debt]):
        wacc = (equity_weight * cost_of_equity) + (debt_weight * after_tax_cost_of_debt); original_wacc = wacc; wacc = min(max(wacc, MIN_WACC), MAX_WACC)
        bound_note = ""
        if abs(wacc - original_wacc) > 0.0001 : bound_note = f"(Bounded from {format_percent(original_wacc)})"; calculation_log.append(f"⚠️ WACC ({format_percent(original_wacc)}) bounded.")
        calculation_log.append(f"▶️ Calculated WACC: {format_percent(wacc)} {bound_note}")
    else: calculation_log.append(f"❌ ERROR: Cannot calculate final WACC due to invalid components."); wacc = np.nan
    components['final_wacc'] = wacc

    display(Markdown("```\n" + "\n".join(calculation_log) + "\n```"))

    if np.isfinite(wacc) and wacc > 0 and np.isfinite(equity_weight) and np.isfinite(debt_weight):
        equity_contrib = equity_weight * cost_of_equity; debt_contrib = debt_weight * after_tax_cost_of_debt
        if np.isfinite(equity_contrib) and np.isfinite(debt_contrib) and abs(equity_contrib + debt_contrib - wacc) < 0.01:
           labels = ['Cost of Equity Contribution', 'Cost of Debt Contribution']; values = [equity_contrib, debt_contrib]
           fig_wacc_donut = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4, marker_colors=px.colors.qualitative.Pastel, hovertemplate='<b>%{label}</b><br>Contribution: %{value:.2%}<extra></extra>', textinfo='percent', textfont_size=12, sort=False)])
           fig_wacc_donut.update_layout(title_text=f'WACC Composition (Total: {format_percent(wacc)})', title_x=0.5, showlegend=True, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), margin=dict(l=20, r=20, t=60, b=50), height=350)
           fig_wacc_donut.show()
        else: display(Markdown(f"*(WACC component contributions invalid/don't sum. Cannot generate chart.)*"))
    elif not np.isfinite(wacc): display(Markdown(f"*(WACC calculation failed. Cannot generate chart)*"))
    else: display(Markdown(f"*(WACC zero/negative or weights invalid. Cannot generate chart)*"))

    return wacc, components

# Execute WACC Calculation
discount_rate, wacc_components = calculate_wacc(
    stock_data['info'], stock_data['financials'], stock_data['balance_sheet'],
    risk_free_rate_input, market_risk_premium_input, beta_input, input_sources,
    currency_symbol, is_indian_stock
)

# %% [markdown]
# ## 6. FCFF Calculation & Projection (Initial Growth: Robust GM of valid 3yr CAGRs)

# %%
# FCFF Function Definition (NEW initial growth logic - V13 - Robust Stock CAGR)
def calculate_and_project_fcff(financials_df, balance_sheet_df, cashflow_df, history_long_df, # History needed for stock CAGR
                               wacc_comps, disc_rate, forecast_yrs,
                               final_user_rf, default_init_growth, # This is 6.5% default
                               currency_sym, is_indian): # Added currency params
    display(Markdown("### Calculating & Projecting FCFF..."))
    projection_log = []
    projection_results = {
        'historical_fcffs': {}, # Store recent historical FCFFs
        'historical_revenues': {}, # Store recent historical revenues
        'current_fcff': np.nan, 'projected_fcffs': [], 'growth_rates': [],
        'pv_fcffs': [], 'cumulative_pv_fcff': np.nan, 'terminal_growth_rate': np.nan,
        'terminal_value': np.nan, 'pv_terminal_value': np.nan, 'error': None,
        'initial_growth_rate': np.nan, 'growth_source': 'N/A'
    }

    if financials_df.empty or balance_sheet_df.empty or cashflow_df.empty:
         projection_results['error'] = "Missing Financials/Balance Sheet/Cash Flow."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results
    if not np.isfinite(disc_rate):
         projection_results['error'] = "Invalid WACC provided."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results

    # --- Define Helper to calculate FCFF for a specific year column ---
    def get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, tax_rt):
        if not isinstance(fin_col, pd.Series) or not isinstance(bs_col, pd.Series) or not isinstance(cf_col, pd.Series): return np.nan, "Missing Statement Column"
        ebit = safe_get(fin_col, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
        depreciation_amortization = safe_get(cf_col, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
        capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
        capital_expenditures = abs(safe_get(cf_col, capex_keys, default=0))
        wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]
        change_in_wc = safe_get(cf_col, wc_keys, default=None) ; wc_src = "CF"
        if change_in_wc is None or not np.isfinite(change_in_wc):
            wc_src = "BS Calc"
            if isinstance(bs_prev_col, pd.Series):
                 curr_assets = safe_get(bs_col, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(bs_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                 prev_assets = safe_get(bs_prev_col, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(bs_prev_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                 if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val): change_in_wc = curr_wc_val - prev_wc_val
                 else: change_in_wc = 0; wc_src = "BS Fail"
            else: change_in_wc = 0; wc_src = "No Prev BS"
        if not np.isfinite(change_in_wc): change_in_wc = 0
        fcff_val = np.nan; calc_status = "OK"
        if np.isfinite(ebit) and np.isfinite(tax_rt):
            nopat_val = ebit * (1 - tax_rt)
            if np.isfinite(depreciation_amortization) and np.isfinite(capital_expenditures) and np.isfinite(change_in_wc): fcff_val = nopat_val + depreciation_amortization - capital_expenditures - change_in_wc
            else: calc_status = "Invalid Depr/CapEx/WC"
        else: calc_status = "Invalid EBIT/Tax"
        return fcff_val, calc_status
    # --- End FCFF Helper ---

    # --- Calculate Historical FCFF & Revenue (need T-0 and T-3 for CAGR) ---
    num_years_cagr = 3
    max_hist_years = num_years_cagr + 1 # Need 4 data points for 3yr CAGR
    num_statement_years = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    historical_fcffs_list = [] # Newest first (index 0 = T-0)
    historical_revenues_list = [] # Newest first
    effective_tax_rate = wacc_comps.get('tax_rate', DEFAULT_TAX_RATE)

    projection_log.append(f"Calculating historical FCFF & Revenue (up to {max_hist_years} points)...")
    years_to_calc = min(num_statement_years, max_hist_years)
    for i in range(years_to_calc):
        fin_col = financials_df.iloc[:, i]; bs_col = balance_sheet_df.iloc[:, i]; cf_col = cashflow_df.iloc[:, i]
        bs_prev_col = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None
        fcff_val, status = get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, effective_tax_rate)
        historical_fcffs_list.append(fcff_val if status == "OK" else np.nan)
        rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
        revenue_val = safe_get(fin_col, rev_keys, default=np.nan)
        historical_revenues_list.append(revenue_val)

    projection_results['historical_fcffs'] = {f"T-{i}": v for i, v in enumerate(historical_fcffs_list)}
    projection_results['historical_revenues'] = {f"T-{i}": v for i, v in enumerate(historical_revenues_list)}
    fcff_current = historical_fcffs_list[0] if historical_fcffs_list and np.isfinite(historical_fcffs_list[0]) else np.nan
    projection_results['current_fcff'] = fcff_current

    # --- Display Calculation Log for Current FCFF (if calculated) ---
    if np.isfinite(fcff_current):
       latest_fin = financials_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]
       ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
       nopat = ebit * (1-effective_tax_rate) if np.isfinite(ebit) else np.nan
       depreciation_amortization = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
       capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]; capital_expenditures = abs(safe_get(latest_cf, capex_keys, default=0))
       wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]; change_in_wc = safe_get(latest_cf, wc_keys, default=None); wc_source = "Direct from Cash Flow"
       if change_in_wc is None or not np.isfinite(change_in_wc):
            if balance_sheet_df.shape[1] >= 2:
                prev_bs = balance_sheet_df.iloc[:, 1]
                curr_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0); curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                prev_assets = safe_get(prev_bs, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(prev_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0); prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val): change_in_wc = curr_wc_val - prev_wc_val; wc_source = "Calculated from BS"
                else: change_in_wc = 0; wc_source = "Calc Failed (Invalid BS)"
            else: change_in_wc = 0; wc_source = "Default (No Prev BS)"
       if not np.isfinite(change_in_wc): change_in_wc = 0
       projection_log.append(f"EBIT (T-0): {format_currency(ebit, currency_sym, is_indian=is_indian)}, Tax Rate ({wacc_comps.get('tax_rate_source','?')}) {format_percent(effective_tax_rate)} -> NOPAT: {format_currency(nopat, currency_sym, is_indian=is_indian)}")
       projection_log.append(f"  + Dep&Amort: {format_currency(depreciation_amortization, currency_sym, is_indian=is_indian)}")
       projection_log.append(f"  - CapEx: {format_currency(capital_expenditures, currency_sym, is_indian=is_indian)}")
       projection_log.append(f"  - Change WC ({wc_source}): {format_currency(change_in_wc, currency_sym, is_indian=is_indian)}")
       projection_log.append(f"▶️ Historical FCFF (T-0): {format_currency(fcff_current, currency_sym, is_indian=is_indian)}")
    else:
        projection_log.append("⚠️ Could not calculate current historical FCFF (T-0).")

    # Exit if T-0 FCFF is invalid, as it's needed for projection start.
    if not np.isfinite(fcff_current):
        projection_results['error'] = "Invalid historical FCFF (T-0). Cannot proceed.";
        display(Markdown(f"❌ ERROR: {projection_results['error']}")); display(Markdown("```\n" + "\n".join(projection_log) + "\n```")); return projection_results

    # --- REFINED Initial Growth Rate Calculation (GM of valid 3yr CAGRs - V13) ---
    projection_log.append("--- Determining Initial Growth Rate (Geometric Mean of valid 3yr CAGRs) ---")
    initial_growth_rate = np.nan # Start with NaN
    growth_source = "N/A"
    calculated_cagrs = {} # Store individual CAGRs {Name: value}
    valid_cagr_list = [] # Store only the valid CAGRs for GM calculation

    # 1. Calculate 3-Year Revenue CAGR (Conditional)
    rev_cagr = np.nan
    projection_log.append(f"Calculating Revenue {num_years_cagr}yr CAGR...")
    if len(historical_revenues_list) >= max_hist_years:
        rev_end = historical_revenues_list[0]           # T-0
        rev_start = historical_revenues_list[num_years_cagr] # T-3
        # Check data validity BEFORE calculating CAGR
        if pd.notna(rev_end) and math.isfinite(rev_end) and pd.notna(rev_start) and math.isfinite(rev_start) and rev_start > 1e-9:
            rev_cagr = calculate_cagr_modified(rev_start, rev_end, num_years_cagr)
            if pd.notna(rev_cagr) and math.isfinite(rev_cagr):
                 calculated_cagrs['Revenue'] = rev_cagr
                 valid_cagr_list.append(rev_cagr)
                 projection_log.append(f"  ✅ Revenue CAGR: {format_percent(rev_cagr)} (Start: {format_currency(rev_start, '', 0, is_indian)}, End: {format_currency(rev_end, '', 0, is_indian)})")
            else:
                 projection_log.append(f"  ❌ Revenue CAGR Calculation Failed (Result Invalid). (Start: {format_currency(rev_start, '', 0, is_indian)}, End: {format_currency(rev_end, '', 0, is_indian)})")
        else:
            projection_log.append(f"  ❌ Revenue CAGR Skipped (Invalid/Non-positive Start Data). (Start: {format_currency(rev_start, '', 0, is_indian)}, End: {format_currency(rev_end, '', 0, is_indian)})")
    else:
        projection_log.append(f"  ❌ Revenue CAGR Skipped (Insufficient data: {len(historical_revenues_list)} points)")

    # 2. Calculate 3-Year FCFF CAGR (Conditional)
    fcff_cagr = np.nan
    projection_log.append(f"Calculating FCFF {num_years_cagr}yr CAGR...")
    if len(historical_fcffs_list) >= max_hist_years:
        fcff_end = historical_fcffs_list[0]           # T-0
        fcff_start = historical_fcffs_list[num_years_cagr] # T-3
        # Check data validity BEFORE calculating CAGR
        if pd.notna(fcff_end) and math.isfinite(fcff_end) and pd.notna(fcff_start) and math.isfinite(fcff_start) and fcff_start > 1e-9:
            fcff_cagr = calculate_cagr_modified(fcff_start, fcff_end, num_years_cagr)
            if pd.notna(fcff_cagr) and math.isfinite(fcff_cagr):
                 calculated_cagrs['FCFF'] = fcff_cagr
                 valid_cagr_list.append(fcff_cagr)
                 projection_log.append(f"  ✅ FCFF CAGR:    {format_percent(fcff_cagr)} (Start: {format_currency(fcff_start, '', 0, is_indian)}, End: {format_currency(fcff_end, '', 0, is_indian)})")
            else:
                 projection_log.append(f"  ❌ FCFF CAGR Calculation Failed (Result Invalid). (Start: {format_currency(fcff_start, '', 0, is_indian)}, End: {format_currency(fcff_end, '', 0, is_indian)})")
        else:
            projection_log.append(f"  ❌ FCFF CAGR Skipped (Invalid/Non-positive Start Data). (Start: {format_currency(fcff_start, '', 0, is_indian)}, End: {format_currency(fcff_end, '', 0, is_indian)})")
    else:
        projection_log.append(f"  ❌ FCFF CAGR Skipped (Insufficient data: {len(historical_fcffs_list)} points)")

    # 3. Calculate 3-Year Stock Price CAGR (Conditional - Robust Column V13)
    stock_cagr = np.nan
    price_col = None # Determine which price column to use
    projection_log.append(f"Calculating Stock Price {num_years_cagr}yr CAGR...")
    # Check if history_long_df exists, is not empty, has 'Adj Close' or 'Close', and has at least 2 points
    if history_long_df is not None and not history_long_df.empty and len(history_long_df) >= 2:
         if 'Adj Close' in history_long_df.columns:
             price_col = 'Adj Close'
         elif 'Close' in history_long_df.columns:
             price_col = 'Close'
             projection_log.append("  ⚠️ Using 'Close' price for Stock CAGR ('Adj Close' not found).")
         else:
             projection_log.append(f"  ❌ Stock CAGR Skipped (Neither 'Adj Close' nor 'Close' columns found in history data)")

         # Proceed only if a valid price column was found and index is datetime
         if price_col and isinstance(history_long_df.index, pd.DatetimeIndex):
             projection_log.append(f"  Stock History Info: Using '{price_col}' column. Found {len(history_long_df)} rows, Date Range: {history_long_df.index.min().date()} to {history_long_df.index.max().date()}")
             try:
                 end_date = history_long_df.index[-1]
                 start_date_target = end_date - pd.DateOffset(years=num_years_cagr)
                 end_idx = len(history_long_df) - 1
                 start_idx = history_long_df.index.searchsorted(start_date_target, side='left')

                 if start_idx >= len(history_long_df):
                      start_idx = len(history_long_df) - 1 # Use the last available point

                 projection_log.append(f"  Stock CAGR Lookup: Target Start Date={start_date_target.date()}, Found Start Idx={start_idx}, End Idx={end_idx}")

                 if start_idx >= 0 and start_idx < end_idx:
                    start_price = history_long_df[price_col].iloc[start_idx]
                    end_price = history_long_df[price_col].iloc[end_idx]
                    start_actual_date = history_long_df.index[start_idx]
                    end_actual_date = history_long_df.index[end_idx]
                    actual_years = (end_actual_date - start_actual_date).days / 365.25

                    projection_log.append(f"  Stock CAGR Data Points: Start Date={start_actual_date.date()}, Start Price={start_price:.2f}, End Date={end_actual_date.date()}, End Price={end_price:.2f}, Actual Years={actual_years:.2f}")

                    is_end_valid = pd.notna(end_price) and math.isfinite(end_price)
                    is_start_valid = pd.notna(start_price) and math.isfinite(start_price) and start_price > 1e-9
                    is_period_valid = actual_years >= (num_years_cagr - 0.5) and actual_years > 0

                    if is_end_valid and is_start_valid and is_period_valid:
                        stock_cagr = calculate_cagr_modified(start_price, end_price, actual_years)

                        if pd.notna(stock_cagr) and math.isfinite(stock_cagr):
                            calculated_cagrs['Stock'] = stock_cagr
                            valid_cagr_list.append(stock_cagr)
                            projection_log.append(f"  ✅ Stock CAGR:   {format_percent(stock_cagr)} (Using '{price_col}')")
                        else:
                            projection_log.append(f"  ❌ Stock CAGR Calculation Failed (Result Invalid).")
                    else:
                        reason = []
                        if not is_end_valid: reason.append("Invalid End Price")
                        if not is_start_valid: reason.append("Invalid/Non-positive Start Price")
                        if not is_period_valid: reason.append(f"Period Invalid ({actual_years:.2f} yrs vs target {num_years_cagr})")
                        projection_log.append(f"  ❌ Stock CAGR Skipped ({', '.join(reason)}).")
                 else:
                      projection_log.append(f"  ❌ Stock CAGR Skipped (Could not find valid start index < end index: StartIdx={start_idx}, EndIdx={end_idx})")
             except IndexError as e:
                  projection_log.append(f"  ❌ Stock CAGR Skipped (IndexError during data lookup - start_idx={start_idx}, end_idx={end_idx}, Error: {e})")
             except Exception as e:
                  projection_log.append(f"  ❌ Stock CAGR Skipped (Error during lookup/calculation: {e})")
         elif not price_col:
             # Already logged the reason for no price column
             pass
         else: # Index is not datetime
             projection_log.append(f"  ❌ Stock CAGR Skipped (History index is not DatetimeIndex)")

    else: # history_long_df checks failed
         reason = []
         if history_long_df is None: reason.append("History DF is None")
         elif history_long_df.empty: reason.append("History DF is empty")
         elif len(history_long_df) < 2: reason.append("Less than 2 data points")
         projection_log.append(f"  ❌ Stock CAGR Skipped (Insufficient/invalid history data object: {', '.join(reason)})")


    # 4. Calculate Geometric Mean of the valid CAGRs
    if valid_cagr_list:
        gm_growth = geometric_mean_modified(valid_cagr_list)
        if pd.notna(gm_growth) and math.isfinite(gm_growth):
            initial_growth_rate = gm_growth
            growth_source = f"GM of {len(valid_cagr_list)} valid CAGR(s)"
            projection_log.append(f"  ▶️ Calculated Geometric Mean = {format_percent(initial_growth_rate)}")
            included_cagrs = [f"{k}:{format_percent(v)}" for k, v in calculated_cagrs.items() if pd.notna(v) and math.isfinite(v)] # Rebuild list from dict
            projection_log.append(f"     Based on: [{', '.join(included_cagrs)}]")
        else:
            projection_log.append(f"  ⚠️ Geometric Mean calculation failed or result invalid. Falling back to default.")
            initial_growth_rate = default_init_growth
            growth_source = f"Default ({format_percent(default_init_growth)}) [GM Fail]"
            projection_log.append(f"     CAGRs calculated but GM failed: {valid_cagr_list}") # Log the rates that caused GM failure
    else:
        # If NO valid CAGRs were calculated
        projection_log.append(f"  ⚠️ No valid 3yr CAGRs could be calculated. Falling back to default.")
        initial_growth_rate = default_init_growth
        growth_source = f"Default ({format_percent(default_init_growth)}) [No Valid CAGRs]"

    # 5. Apply Floor (1.1%) and Ceiling (15%)
    unbounded_rate = initial_growth_rate # Store original for comparison
    initial_growth_rate = max(initial_growth_rate, MIN_INITIAL_GROWTH_FLOOR) # Apply 1.1% floor
    initial_growth_rate = min(initial_growth_rate, MAX_INITIAL_GROWTH_CEILING) # Apply 15% ceiling

    if abs(initial_growth_rate - unbounded_rate) > 1e-6: # Check if bounding changed the value
        growth_source += " (Bounded)"
        projection_log.append(f"  Applied Bounds: Final Initial Growth Rate = {format_percent(initial_growth_rate)} (Floor: {format_percent(MIN_INITIAL_GROWTH_FLOOR)}, Ceiling: {format_percent(MAX_INITIAL_GROWTH_CEILING)})")
    else:
         projection_log.append(f"  Final Initial Growth Rate = {format_percent(initial_growth_rate)}")

    projection_results['initial_growth_rate'] = initial_growth_rate
    projection_results['growth_source'] = growth_source

    # --- Terminal Growth Rate Calculation (Remains the same conservative approach) ---
    g_cap_initial_valid = initial_growth_rate if pd.notna(initial_growth_rate) and math.isfinite(initial_growth_rate) else TERMINAL_GROWTH_RATE_CAP # Fallback if initial is invalid
    g_cap_config = TERMINAL_GROWTH_RATE_CAP; g_cap_wacc = disc_rate * 0.67; g_cap_rf = final_user_rf * 0.95 # Use final RF
    projection_log.append(f"--- Determining Terminal Growth Rate ---")
    projection_log.append(f"Terminal Growth Caps Considered: Config={format_percent(g_cap_config)}, WACC*0.67={format_percent(g_cap_wacc)}, Initial(Final)={format_percent(g_cap_initial_valid)}, Rf*0.95={format_percent(g_cap_rf)}")
    terminal_growth_rate = min(g_cap_config, g_cap_wacc, g_cap_initial_valid, g_cap_rf); projection_log.append(f"↳ Minimum of Caps: {format_percent(terminal_growth_rate)}")
    original_g_before_floor = terminal_growth_rate
    terminal_growth_rate = max(MIN_GROWTH, terminal_growth_rate) # Apply floor (MIN_GROWTH=0.00 usually)
    if abs(terminal_growth_rate - original_g_before_floor) > 1e-6 : projection_log.append(f"↳ Adjusted by Floor ({format_percent(MIN_GROWTH)}): {format_percent(terminal_growth_rate)}")
    if terminal_growth_rate >= disc_rate: original_g = terminal_growth_rate; terminal_growth_rate = disc_rate - 0.005; projection_log.append(f"⚠️ Terminal Growth ({format_percent(original_g)}) >= WACC ({format_percent(disc_rate)}). Reduced to {format_percent(terminal_growth_rate)}.")
    projection_log.append(f"▶️ Using Final Terminal Growth Rate (g): {format_percent(terminal_growth_rate)}"); projection_results['terminal_growth_rate'] = terminal_growth_rate

    # --- Project FCFFs using Linear Decay from initial_growth_rate -> terminal_growth_rate ---
    projected_fcffs = []; growth_rates_applied = []; pv_fcffs = []; cumulative_pv_fcff = 0.0
    current_fcff_for_proj = fcff_current # Start projection base from T-0 FCFF

    if pd.isna(initial_growth_rate) or not math.isfinite(initial_growth_rate):
        projection_log.append(f"⚠️ FATAL: Invalid final initial growth rate ({initial_growth_rate}) despite checks. Using 0% for projection start.")
        effective_initial_growth = 0.0
    else:
        effective_initial_growth = initial_growth_rate # Use the final bounded rate

    effective_terminal_growth = terminal_growth_rate

    if current_fcff_for_proj <= 0:
        projection_log.append("⚠️ Starting FCFF (T-0) <= 0. Applying minimal growth.")
        effective_initial_growth = min(effective_initial_growth, 0.01);
        effective_terminal_growth = min(terminal_growth_rate, 0.00)

    projection_log.append("--- FCFF Projections ---"); projection_log.append(f"{'Year':<5} {'Growth':<8} {'FCFF':<18} {'PV Factor':<10} {'PV FCFF':<18}"); projection_log.append("-" * 60)
    last_valid_fcff = current_fcff_for_proj

    for year in range(1, forecast_yrs + 1):
        if forecast_yrs > 0: decay_factor = max(0, (forecast_yrs - year)) / forecast_yrs
        else: decay_factor = 0
        growth_rate = effective_terminal_growth + (effective_initial_growth - effective_terminal_growth) * decay_factor
        growth_rate = max(MIN_GROWTH, growth_rate) # Apply general floor (0%)
        if current_fcff_for_proj <= 0: growth_rate = max(-0.10, min(growth_rate, 0.05)) # Moderate if start neg

        growth_rates_applied.append(growth_rate)
        fcff_forecast = last_valid_fcff * (1 + growth_rate)
        if not np.isfinite(fcff_forecast): fcff_forecast = 0
        projected_fcffs.append(fcff_forecast)
        last_valid_fcff = fcff_forecast # Update last_valid_fcff

        pv_factor = 1 / ((1 + disc_rate) ** year); pv_fcff = fcff_forecast * pv_factor
        if not np.isfinite(pv_fcff): pv_fcff = 0
        pv_fcffs.append(pv_fcff); cumulative_pv_fcff += pv_fcff
        projection_log.append(f"{year:<5} {format_percent(growth_rate):<8} {format_currency(fcff_forecast, '', 0, is_indian=is_indian):<18} {pv_factor:<10.4f} {format_currency(pv_fcff, '', 0, is_indian=is_indian):<18}")

    projection_results['projected_fcffs'] = projected_fcffs; projection_results['growth_rates'] = growth_rates_applied; projection_results['pv_fcffs'] = pv_fcffs
    projection_results['cumulative_pv_fcff'] = cumulative_pv_fcff if np.isfinite(cumulative_pv_fcff) else 0.0
    projection_log.append("-" * 60); projection_log.append(f"Sum of PV(Forecasted FCFFs): {format_currency(cumulative_pv_fcff, symbol=currency_sym, is_indian=is_indian)}")

    # Calculate Terminal Value (Remains the same)
    terminal_value = np.nan; pv_terminal_value = np.nan
    final_forecast_fcff = projected_fcffs[-1] if projected_fcffs and np.isfinite(projected_fcffs[-1]) else 0

    if (disc_rate - terminal_growth_rate) <= 1e-6 or not np.isfinite(terminal_growth_rate):
         projection_log.append(f"❌ ERROR: WACC ({format_percent(disc_rate)}) <= Terminal Growth ({format_percent(terminal_growth_rate)}) or g invalid. TV calc failed."); projection_results['error'] = "WACC <= g or g invalid. TV Calc failed."
    elif not np.isfinite(final_forecast_fcff):
         projection_log.append(f"❌ ERROR: Final projected FCFF (Year {forecast_yrs}) is invalid. TV calc failed."); projection_results['error'] = "Final FCFF invalid. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    else:
         if final_forecast_fcff <= 0:
             projection_log.append(f"ℹ️ FCFF Year {forecast_yrs} ({format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)}) <= 0. Setting TV=0.");
             terminal_value = 0.0; pv_terminal_value = 0.0
         else:
              fcff_n_plus_1 = final_forecast_fcff * (1 + terminal_growth_rate)
              terminal_value = fcff_n_plus_1 / (disc_rate - terminal_growth_rate)
              if not np.isfinite(terminal_value) or terminal_value < 0:
                  projection_log.append(f"⚠️ Calculated TV invalid/negative ({format_currency(terminal_value, currency_sym, is_indian=is_indian)}). Setting TV=0.")
                  terminal_value = 0.0; pv_terminal_value = 0.0
              else:
                   pv_factor_tv = 1 / ((1 + disc_rate) ** forecast_yrs); pv_terminal_value = terminal_value * pv_factor_tv
                   projection_log.append(f"FCFF Yr {forecast_yrs}: {format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)} -> FCFF Yr {forecast_yrs+1}: {format_currency(fcff_n_plus_1, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"Terminal Value (Yr {forecast_yrs}): {format_currency(terminal_value, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"PV Factor TV: {pv_factor_tv:.4f} -> PV(TV): {format_currency(pv_terminal_value, currency_sym, is_indian=is_indian)}")

    projection_results['terminal_value'] = terminal_value if np.isfinite(terminal_value) else 0.0
    projection_results['pv_terminal_value'] = pv_terminal_value if np.isfinite(pv_terminal_value) else 0.0

    display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
    return projection_results

# --- Execute FCFF calculation and projection ---
# Initialize fcff_results dictionary to ensure it's defined before the try block
fcff_results = {'error': 'Calculation not run yet.', 'initial_growth_rate': np.nan, 'growth_source':'N/A', 'terminal_growth_rate': np.nan}

if np.isfinite(discount_rate):
    try:
        # Pass currency symbol, flag, default growth rate, AND long history
        fcff_results = calculate_and_project_fcff(
            stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'], stock_data['history_long'], # Pass history_long
            wacc_components, discount_rate, FORECAST_YEARS, risk_free_rate_input, DEFAULT_GROWTH_RATE,
            currency_symbol, is_indian_stock)
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during FCFF Projection: {e}**"))
         # Ensure fcff_results is still a dict with error info
         fcff_results = {'error': f'Runtime error in FCFF calc: {e}', 'initial_growth_rate': np.nan, 'growth_source':'Error', 'terminal_growth_rate': np.nan}

else:
    display(Markdown("❌ **Skipping FCFF Projection due to invalid WACC.**"));
    # Ensure fcff_results is a dict even if skipped
    fcff_results = {'error': 'Invalid WACC.', 'initial_growth_rate': np.nan, 'growth_source':'N/A (WACC Invalid)', 'terminal_growth_rate': np.nan}


# %% [markdown]
# ## 7. Enterprise Value and Equity Value Calculation

# %%
# Valuation Function Definition (No changes here)
def calculate_final_valuation(fcff_data, balance_sheet_df, shares_outstanding_val, currency_sym, wacc_data, is_indian): # Added flag
    display(Markdown("### Final Valuation Calculation..."))
    valuation = {'enterprise_value': np.nan, 'equity_value': np.nan, 'intrinsic_value_per_share': np.nan, 'error': None}
    valuation_log = []

    # Ensure fcff_data is a dictionary before accessing keys
    if not isinstance(fcff_data, dict):
        error_msg = f"FCFF Data is not a dictionary (type: {type(fcff_data)}). Cannot proceed."
        valuation['error'] = error_msg
        valuation_log.append(f"❌ ERROR: {error_msg}")
        display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
        return valuation

    if fcff_data.get('error'):
        valuation['error'] = f"FCFF Error: {fcff_data.get('error', 'Unknown')}"; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation

    pv_forecast_sum = fcff_data.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_data.get('pv_terminal_value', np.nan)
    if not np.isfinite(pv_forecast_sum) or not np.isfinite(pv_tv):
         valuation['error'] = f"Invalid PV components (PV FCFF: {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}, PV TV: {format_currency(pv_tv, currency_sym, is_indian=is_indian)})."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation # Pass flag

    enterprise_value = pv_forecast_sum + pv_tv; valuation['enterprise_value'] = enterprise_value
    valuation_log.append(f"PV(Forecast FCFFs): {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"PV(Terminal Value):   {format_currency(pv_tv, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"↳ Enterprise Value (EV): {format_currency(enterprise_value, currency_sym, is_indian=is_indian)}")

    # Get Debt & Cash from BS (use Debt from WACC for consistency)
    total_debt_latest = 0.0; cash_and_equivalents = 0.0; minority_interest = 0.0
    # Ensure wacc_data exists and is a dict before accessing
    debt_source_info = '?'
    if isinstance(wacc_data, dict):
        debt_source_info = wacc_data.get('debt_calc_method', '?')
        total_debt_latest = wacc_data.get('debt_value_used', 0.0) # Get debt used in WACC
    else:
        valuation_log.append(f"⚠️ WACC data invalid, cannot get debt used in WACC. Assuming 0.")

    if not balance_sheet_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        cash_keys = ["Cash And Cash Equivalents", "Cash Cash Equivalents And Short Term Investments", "Cash Financial", "Cash", "Cash And Short Term Investments"]
        cash_and_equivalents = safe_get(latest_bs, cash_keys, "BS", default=0)
        valuation_log.append(f"  + Cash & Equiv (BS): {format_currency(cash_and_equivalents, currency_sym, is_indian=is_indian)}")
        minority_interest = safe_get(latest_bs, ["Minority Interest", "Noncontrolling Interest"], "BS", default=0)
        if minority_interest != 0.0:
            valuation_log.append(f"  - Minority Interest: {format_currency(minority_interest, currency_sym, is_indian=is_indian)}")
    else:
        valuation_log.append("⚠️ Balance Sheet empty.")
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        valuation_log.append(f"  + Cash & Equiv: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")

    if not all(np.isfinite(v) for v in [enterprise_value, total_debt_latest, cash_and_equivalents, minority_interest]):
         valuation['error'] = "Invalid component for EV->Equity bridge."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    equity_value = enterprise_value - total_debt_latest + cash_and_equivalents - minority_interest; valuation['equity_value'] = equity_value
    valuation_log.append(f"↳ Equity Value: {format_currency(equity_value, currency_sym, is_indian=is_indian)}")

    # Calculate Intrinsic Value Per Share
    intrinsic_value_per_share = np.nan
    if not np.isfinite(equity_value): valuation['error'] = f"Equity value invalid ({equity_value})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    elif shares_outstanding_val <= 0: valuation['error'] = f"Shares outstanding invalid ({shares_outstanding_val})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    else:
        intrinsic_value_per_share = equity_value / shares_outstanding_val;
        valuation_log.append(f"  ÷ Shares Out: {shares_outstanding_val:,.0f}")
        valuation_log.append(f"▶️ Intrinsic Value/Share: {format_currency(intrinsic_value_per_share, currency_sym, is_indian=is_indian)}")
    valuation['intrinsic_value_per_share'] = intrinsic_value_per_share if np.isfinite(intrinsic_value_per_share) else np.nan

    display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
    return valuation

# --- Execute Final Valuation ---
# Initialize final_valuation to ensure it's defined before the try block
final_valuation = {'error': 'Calculation not run yet.', 'intrinsic_value_per_share': np.nan}

# Check if fcff_results and wacc_components are dictionaries and fcff_results has no error
if isinstance(fcff_results, dict) and isinstance(wacc_components, dict) and not fcff_results.get('error'):
    try:
        final_valuation = calculate_final_valuation(
            fcff_results, stock_data['balance_sheet'], shares_outstanding,
            currency_symbol, wacc_components, is_indian_stock
        )
    except Exception as e:
        display(Markdown(f"❌ **CRITICAL ERROR during Final Valuation: {e}**"))
        final_valuation = {'error': f'Runtime error in final valuation: {e}', 'intrinsic_value_per_share': np.nan}
elif isinstance(fcff_results, dict): # If fcff_results has an error
     display(Markdown("❌ **Skipping Final Valuation due to FCFF errors.**"));
     final_valuation = {'error': fcff_results.get('error', 'Upstream FCFF error.'), 'intrinsic_value_per_share': np.nan}
else: # If fcff_results or wacc_components are not dictionaries
     display(Markdown("❌ **Skipping Final Valuation due to critical upstream errors (invalid data structures).**"));
     final_valuation = {'error': 'FCFF or WACC results dictionary not created/valid.', 'intrinsic_value_per_share': np.nan}


# %% [markdown]
# ## 8. Financial Ratio Analysis

# %%
# Ratio Function Definition (No changes here)
def analyze_ratios(stock_info_dict, financials_df, balance_sheet_df, cashflow_df,
                   enterprise_val, current_price_val, shares_outstanding_val,
                   currency_sym, user_benchmark_dict, # Only user inputs
                   benchmark_source_dict, # Dict tracking source per ratio
                   wacc_data, is_indian): # Pass WACC data for consistent debt and flag

    """ Calculates ratios, compares to user/fallback benchmarks, uses WACC-consistent debt. """
    display(Markdown("### Financial Ratio Analysis..."))
    ratios = {} # Calculated ratios {RatioName: value}
    ratio_log = []
    ratio_data_for_table = [] # For HTML table display
    ratio_plotly_data = {'Ratio': [], 'Company': [], 'Benchmark': [], 'Benchmark Source': [], 'Status': []} # For Plotly chart
    company_nm = stock_info_dict.get("shortName", ticker_symbol)

    # --- Pre-checks ---
    valid_price = current_price_val is not None and np.isfinite(current_price_val) and current_price_val > 0
    valid_shares = shares_outstanding_val is not None and np.isfinite(shares_outstanding_val) and shares_outstanding_val > 0
    valid_enterprise_val = enterprise_val is not None and np.isfinite(enterprise_val)

    # Check for essential statements
    if financials_df.empty or balance_sheet_df.empty:
         ratio_log.append("❌ ERROR: Income Statement or Balance Sheet empty. Cannot calculate most ratios.")
         display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
         return {}, [] # Return empty results

    latest_fin = financials_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0] if not cashflow_df.empty else pd.Series(dtype=float)

    # --- Calculate Ratio Components (using safe_get with NaN default where appropriate) ---
    net_income = safe_get(latest_fin, ["Net Income Continuous Operations", "Net Income", "Net Income Common Stockholders", "Net Income Applicable To Common Shares"], default=np.nan)
    revenue = safe_get(latest_fin, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
    depreciation = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0) # Default 0 ok here
    ebitda = (ebit + depreciation) if np.isfinite(ebit) and np.isfinite(depreciation) else np.nan

    total_equity = safe_get(latest_bs, ["Total Stockholder Equity", "Stockholders Equity", "Total Equity Gross Minority Interest", "Common Stock Equity"], default=np.nan)
    total_assets = safe_get(latest_bs, ["Total Assets"], default=np.nan)
    current_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=np.nan)
    current_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=np.nan)
    inventory = safe_get(latest_bs, ["Inventory"], default=0) # Default 0 ok here

    # --- Use Debt value consistent with WACC calculation ---
    total_debt_ratio = 0.0; debt_source_info = '?'
    if isinstance(wacc_data, dict):
        total_debt_ratio = wacc_data.get('debt_value_used', 0.0)
        debt_source_info = wacc_data.get('debt_calc_method', '?');
    else:
        ratio_log.append(f"⚠️ WACC data invalid, cannot get debt used in WACC. Assuming 0 for ratio calcs.")
    ratio_log.append(f"Using Debt = {format_currency(total_debt_ratio, currency_sym, is_indian=is_indian)} (Source: WACC - {debt_source_info})")

    # Per Share Metrics
    eps = (net_income / shares_outstanding_val) if valid_shares and np.isfinite(net_income) and net_income != 0 else np.nan
    bvps = (total_equity / shares_outstanding_val) if valid_shares and np.isfinite(total_equity) and total_equity != 0 else np.nan

    # Capital Employed (ROCE denominator)
    capital_employed = np.nan; ce_source = "N/A"
    if np.isfinite(total_assets) and np.isfinite(current_liab):
        capital_employed_def1 = total_assets - current_liab
        if np.isfinite(capital_employed_def1) and capital_employed_def1 > 1.0:
            capital_employed = capital_employed_def1; ce_source = "A-CL"
    if not np.isfinite(capital_employed):
        if np.isfinite(total_equity) and np.isfinite(total_debt_ratio):
             capital_employed_def2 = total_equity + total_debt_ratio
             if np.isfinite(capital_employed_def2) and capital_employed_def2 > 1.0:
                 capital_employed = capital_employed_def2; ce_source = "E+D"
    if not np.isfinite(capital_employed):
         ce_source = "Invalid/Unavailable"; ratio_log.append("⚠️ Capital Employed could not be reliably calculated. ROCE affected.")

    # --- Calculate Ratios (Result is NaN if components invalid/zero denominator) ---
    ratios['P/E'] = (current_price_val / eps) if valid_price and np.isfinite(eps) and eps != 0 else np.nan
    ratios['P/B'] = (current_price_val / bvps) if valid_price and np.isfinite(bvps) and bvps != 0 else np.nan
    ratios['EV/EBITDA'] = (enterprise_val / ebitda) if valid_enterprise_val and np.isfinite(ebitda) and ebitda != 0 else np.nan
    ratios['ROE'] = (net_income / total_equity) if np.isfinite(net_income) and np.isfinite(total_equity) and total_equity != 0 else np.nan
    ratios['ROA'] = (net_income / total_assets) if np.isfinite(net_income) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['ROCE'] = (ebit / capital_employed) if np.isfinite(ebit) and np.isfinite(capital_employed) and capital_employed != 0 else np.nan
    ratios['Asset Turnover'] = (revenue / total_assets) if np.isfinite(revenue) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['Current Ratio'] = (current_assets / current_liab) if np.isfinite(current_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan
    quick_assets = (current_assets - inventory) if np.isfinite(current_assets) and np.isfinite(inventory) else np.nan
    ratios['Quick Ratio'] = (quick_assets / current_liab) if np.isfinite(quick_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan

    # --- Compare with User/Fallback Benchmarks & Prepare Data Tables ---
    display(Markdown("*(Comparing ratios to user/fallback benchmarks...)*"))
    for key in REQUIRED_BENCHMARK_RATIOS:
        company_value = ratios.get(key); is_percent_type = key in ['ROE', 'ROA', 'ROCE']
        value_fmt = format_percent(company_value, 1) if is_percent_type else format_currency(company_value, '', 2, is_indian=False)
        if value_fmt == "N/A" and company_value is not None and not math.isfinite(company_value): value_fmt = "Inf"

        user_benchmark = user_benchmark_dict.get(key, None); fallback_benchmark = DEFAULT_BENCHMARK_RATIOS.get(key, None)
        benchmark_source = benchmark_source_dict.get(key, "?"); final_benchmark = None
        if benchmark_source == 'User Input' and user_benchmark is not None: final_benchmark = user_benchmark
        elif benchmark_source == 'Fallback Default' and fallback_benchmark is not None: final_benchmark = fallback_benchmark
        else: benchmark_source = "Not Available"

        bench_fmt = "N/A"
        if final_benchmark is not None: bench_fmt = f"{final_benchmark:.1f}%" if is_percent_type else f"{final_benchmark:.2f}"; bench_fmt += f" <i style='font-size:0.8em; color:grey'>({benchmark_source.replace('Fallback ','')})</i>"
        else: bench_fmt = "<i style='color:grey'>Not Provided</i>"

        status = "<i style='color:grey'>N/A</i>"; status_color = "grey"
        if np.isfinite(company_value) and final_benchmark is not None and np.isfinite(final_benchmark):
            numeric_value = company_value; numeric_benchmark = final_benchmark
            good_color = '#2ecc71'; bad_color = '#e74c3c'; neutral_color = '#f39c12'; grey_color='#7f8c8d'
            if abs(numeric_benchmark) > 1e-9: ratio_diff = (numeric_value - numeric_benchmark) / abs(numeric_benchmark)
            else: ratio_diff = float('inf') if numeric_value > 0 else float('-inf') if numeric_value < 0 else 0.0
            if key in ['ROE', 'ROA', 'ROCE', 'Asset Turnover', 'Current Ratio', 'Quick Ratio']: # Higher is better
                 if ratio_diff > 0.20: status_color = good_color; status = f"⬆️ Strong"
                 elif ratio_diff < -0.20: status_color = bad_color; status = f"⬇️ Weak"
                 else: status_color = grey_color; status = f"~ Average"
            elif key in ['P/E', 'P/B', 'EV/EBITDA']: # Lower might be better (valuation) or neutral
                 if ratio_diff < -0.20: status_color = good_color; status = f"✅ Low"
                 elif ratio_diff > 0.20: status_color = neutral_color; status = f"⚠️ High"
                 else: status_color = grey_color; status = f"~ Average"
            status = f"<span style='color:{status_color}'>{status}</span>"
            plot_value = numeric_value * 100 if is_percent_type else numeric_value; plot_benchmark = numeric_benchmark
            ratio_plotly_data['Ratio'].append(key); ratio_plotly_data['Company'].append(plot_value); ratio_plotly_data['Benchmark'].append(plot_benchmark);
            ratio_plotly_data['Benchmark Source'].append(benchmark_source); ratio_plotly_data['Status'].append(status)
        elif not np.isfinite(company_value): status = "<i style='color:grey'>Company N/A</i>"
        elif final_benchmark is None: status = "<i style='color:grey'>No Benchmark</i>"
        ratio_data_for_table.append([f"<b>{key}</b>", value_fmt, bench_fmt, status])

    # --- Display Logs & Table ---
    if ratio_log: display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
    if ratio_data_for_table:
        header = ["Ratio", f"{company_nm} Value", "Benchmark (Source)", "Status"]; html_table = "<p>(*Status vs User/Default benchmarks. Threshold: +/- 20%.)</p>"; html_table += "<table border='1' style='border-collapse:collapse; width:98%; margin-left:auto; margin-right:auto; font-size: 0.9em;'>"; html_table += "<thead style='background-color:#e9ecef; color: #495057;'><tr>" + "".join(f"<th style='padding: 8px; text-align:center;'>{h}</th>" for h in header) + "</tr></thead>"; html_table += "<tbody>"
        for i, row in enumerate(ratio_data_for_table): bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff"; html_table += f"<tr style='background-color:{bg_color};'>"; html_table += f"<td style='padding: 6px; text-align:left;'>{row[0]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[1]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[2]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[3]}</td>"; html_table += "</tr>"
        html_table += "</tbody></table><br>"; display(HTML(html_table))
    else: display(Markdown("*No ratio data available.*"))

    # --- Display Ratio Comparison Chart (Plotly Bullet) ---
    if ratio_plotly_data['Ratio']:
       num_ratios = len(ratio_plotly_data['Ratio']); fig_ratios = go.Figure()
       def get_range_colors(k): return ["#fadbd8", "#fdebd0", "#d5f5e3"] if k in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio'] else ["#d5f5e3","#fdebd0","#fadbd8"] if k in ['P/E','P/B','EV/EBITDA'] else ['lightgrey','darkgrey','grey']
       for i in range(num_ratios):
           ratio_name = ratio_plotly_data['Ratio'][i]; company_val = ratio_plotly_data['Company'][i]; benchmark_val = ratio_plotly_data['Benchmark'][i]; bench_src_plot = ratio_plotly_data['Benchmark Source'][i]
           if not np.isfinite(company_val) or not np.isfinite(benchmark_val): print(f"Skipping plot for {ratio_name}: Invalid Value ({company_val}) or Bench ({benchmark_val})."); continue
           range_colors = get_range_colors(ratio_name); is_percent = ratio_name in ['ROE', 'ROA', 'ROCE']
           poor_f=0.8; good_f=1.2; range1_limit = benchmark_val * (poor_f if benchmark_val >=0 else good_f); range2_limit = benchmark_val * (good_f if benchmark_val >=0 else poor_f)
           abs_max = max(abs(company_val), abs(benchmark_val)); plot_max_val = abs_max * 1.7 + 1; plot_min_val = 0
           if company_val < 0 or benchmark_val < 0: plot_min_val = min(company_val, benchmark_val) * 1.7 - 1
           if ratio_name in ['P/E', 'EV/EBITDA']: plot_max_val = min(plot_max_val, 100); plot_min_val = max(plot_min_val, -20)
           elif ratio_name in ['P/B']: plot_max_val = min(plot_max_val, 20); plot_min_val = max(plot_min_val, -5)
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: plot_max_val = min(plot_max_val, 5); plot_min_val = max(plot_min_val, 0)
           elif is_percent: plot_max_val = min(plot_max_val, 75); plot_min_val= max(plot_min_val,-50)
           else: plot_max_val = min(plot_max_val, 1000)
           if plot_min_val >= plot_max_val: plot_max_val = plot_min_val + 1
           steps = [];
           if abs(benchmark_val) < 1e-9: steps = [{'range': [plot_min_val, plot_max_val], 'color':'lightgrey'}]
           else: step_ranges = sorted([plot_min_val, range1_limit, range2_limit, plot_max_val]); step_colors = range_colors if benchmark_val >= 0 else range_colors[::-1]; steps = [{'range': [step_ranges[j], step_ranges[j+1]], 'color': step_colors[j]} for j in range(3)]
           bench_fmt_title = f"{benchmark_val:.1f}%" if is_percent else f"{benchmark_val:.2f}"; title_text = f"<b>{ratio_name}</b><br><i style='font-size:0.8em'>(Bench: {bench_fmt_title} [{bench_src_plot.replace('Fallback ','')}])</i>"
           lower_is_better = ratio_name in ['P/E', 'P/B', 'EV/EBITDA']; delta_decreasing_color = good_color if lower_is_better else bad_color; delta_increasing_color = bad_color if lower_is_better else good_color
           fig_ratios.add_trace(go.Indicator(
                mode = "number+gauge+delta", value = company_val, number={'suffix': "%" if is_percent else "", 'valueformat': ".1f"},
                delta = {'reference': benchmark_val, 'relative': True, 'valueformat':".0%", 'decreasing': {'color': delta_decreasing_color}, 'increasing': {'color': delta_increasing_color}},
                domain = {'row': i, 'column': 0}, title = {'text' : title_text, 'font':{'size':11}},
                gauge = { 'shape': "bullet", 'axis': {'range': [plot_min_val, plot_max_val] if plot_min_val < plot_max_val else [plot_min_val, plot_min_val+1], 'tickformat':'.1f'},
                    'threshold': { 'line': {'color': "black", 'width': 2}, 'thickness': 0.8, 'value': benchmark_val },
                    'steps': [ s for s in steps if s and isinstance(s.get('range'), list) and len(s['range'])==2 and s['range'][0] < s['range'][1] ],
                    'bar': {'color': "#2c3e50", 'thickness': 0.5} }
            ))
       fig_ratios.update_layout(height = max(300, num_ratios * 85), grid = {'rows': num_ratios, 'columns': 1, 'pattern': "independent"}, title_text='Ratio Performance vs. Benchmarks', title_x=0.5, margin=dict(l=160, r=40, t=60, b=30))
       fig_ratios.show()
    else: display(Markdown("*No valid ratios/benchmarks for comparison chart.*"))

    return ratios, ratio_data_for_table

# --- Execute Ratio Analysis ---
# Initialize ratios_calculated and ratio_table before the try block
ratios_calculated = {}; ratio_table = []

# Check if final_valuation and wacc_components are dictionaries and final_valuation has no error
if isinstance(final_valuation, dict) and isinstance(wacc_components, dict) and not final_valuation.get('error'):
    try:
        # Retrieve enterprise value safely from final_valuation dict
        ev_for_ratios = final_valuation.get('enterprise_value', np.nan)
        ratios_calculated, ratio_table = analyze_ratios(
            stock_data['info'], stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'],
            ev_for_ratios, current_price, shares_outstanding, currency_symbol,
            user_benchmarks, benchmark_sources, wacc_components, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Ratio Analysis: {e}**"))
         # Ensure they are reset on error
         ratios_calculated = {}; ratio_table = []
else:
    display(Markdown("❌ **Skipping Ratio Analysis due to prior errors or invalid data structures.**"));
    # Ensure they are reset if skipped
    ratios_calculated = {}; ratio_table = []


# %% [markdown]
# ## 9. Summary & Visualization Dashboard

# %%
# Dashboard Function Definition (No changes here, relies on valid inputs)
def display_summary_dashboard(company_nm, ticker_sym, current_pr, intrinsic_val_per_sh,
                              wacc_val, term_growth, init_growth, growth_src,
                              final_val_dict, hist_df, fcff_res, currency_sym, info_dict, input_src_dict,
                              is_indian): # Added flag

    display(Markdown(f"## Valuation Summary Dashboard: {company_nm} ({ticker_sym})"))
    source_summary = " | ".join(f"{k}:`{v.split('(')[0].strip()}`" for k, v in input_src_dict.items())
    growth_src_display = growth_src if growth_src else 'N/A'
    display(Markdown(f"**Input Sources:** {source_summary} | Growth:`{growth_src_display}`"))

    valid_intrinsic = intrinsic_val_per_sh is not None and np.isfinite(intrinsic_val_per_sh)
    valid_current_pr = current_pr is not None and np.isfinite(current_pr) and current_pr > 0
    valid_wacc = wacc_val is not None and np.isfinite(wacc_val)
    valid_term_g = term_growth is not None and np.isfinite(term_growth)
    valid_init_g = init_growth is not None and np.isfinite(init_growth)

    # --- Indicator Subplots ---
    fig_indicators = make_subplots(rows=1, cols=3, specs=[[{'type': 'indicator'}, {'type': 'indicator'}, {'type': 'indicator'}]],
                                 subplot_titles=("Price vs. Intrinsic", "WACC", "Growth Rates"), horizontal_spacing=0.1, column_widths=[0.4, 0.3, 0.3])

    # Indicator 1: Price vs Intrinsic
    intrinsic_val_num = intrinsic_val_per_sh if valid_intrinsic else np.nan
    intrinsic_val_fmt = format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian) if valid_intrinsic else "N/A"
    current_pr_num = current_pr if valid_current_pr else 0.0; upside_fmt = ""; delta_config = None
    if valid_intrinsic and valid_current_pr :
        if intrinsic_val_num > 0: upside_pct = (intrinsic_val_num / current_pr_num) - 1; upside_fmt = f" ({upside_pct:+.1%})"
        comparison_text = f"Intrinsic: {intrinsic_val_fmt}{upside_fmt}"
        delta_config = {'reference': intrinsic_val_num, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym, 'decreasing': {'color': '#2ecc71', 'symbol': '▲ Upside'}, 'increasing': {'color': '#e74c3c', 'symbol': '▼ Downside'}}
    elif valid_intrinsic: comparison_text = f"Intrinsic: {intrinsic_val_fmt} (Current N/A)"
    else: comparison_text = f"Intrinsic: N/A"
    fig_indicators.add_trace(go.Indicator(mode = "number+delta" if delta_config else "number", value = current_pr_num if valid_current_pr else 0, number={'prefix': currency_sym + " " if is_indian else currency_sym, 'valueformat': '.2f', 'font':{'size':24}}, title = {"text": f"<span style='font-size:0.8em'>{comparison_text}</span>"}, delta = delta_config, domain = {'row': 0, 'column': 0}), row=1, col=1)

    # Indicator 2: WACC Gauge
    wacc_val_num = wacc_val * 100 if valid_wacc else None
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = wacc_val_num,
        number = {'suffix': "%", 'valueformat':'.1f' if valid_wacc else '', 'font':{'size':20}},
        gauge = {'axis': {'range': [MIN_WACC*100, MAX_WACC*100]}, 'bar': {'color': "darkblue" if valid_wacc else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc", 'steps' : [ {'range': [MIN_WACC*100, 9], 'color': "#d5f5e3"}, {'range': [9, 14], 'color': "#fdebd0"}, {'range': [14, MAX_WACC*100], 'color': "#fadbd8"}]} if valid_wacc else None,
        title = {"text": "<span style='font-size:0.8em'>WACC</span>"}, domain = {'row': 0, 'column': 1}), row=1, col=2)

    # Indicator 3: Growth Gauge
    term_growth_num = term_growth * 100 if valid_term_g else None; init_growth_fmt = format_percent(init_growth, 1) if valid_init_g else "N/A"
    growth_title_text = f"<span style='font-size:0.8em'>Terminal Growth (g)<br>(Initial: {init_growth_fmt})</span>"
    gauge_max_range = max(TERMINAL_GROWTH_RATE_CAP * 100 + 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) + 1)
    gauge_min_range = MIN_GROWTH * 100 - 1
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = term_growth_num,
         number = {'suffix': "%", 'valueformat':'.1f' if valid_term_g else '', 'font':{'size':20}},
         gauge = {'axis': {'range': [gauge_min_range, gauge_max_range], 'tickwidth': 1}, 'bar': {'color': "darkgrey" if valid_term_g else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [{'range': [gauge_min_range, 2.0], 'color': "lightcoral"}, {'range': [2.0, 4.5], 'color': "lightyellow"}, {'range': [4.5, gauge_max_range], 'color': "lightgreen"}],
                 'threshold' : {'line': {'color': "black", 'width': 3}, 'thickness': 0.75, 'value': term_growth_num } if valid_term_g else None } if valid_term_g else None,
         title = {"text": growth_title_text}, domain = {'row': 0, 'column': 2}), row=1, col=3)

    fig_indicators.update_layout(height=220, margin=dict(l=30, r=30, t=50, b=20)); fig_indicators.show()

    # --- Historical Price Chart ---
    if not hist_df.empty:
        fig_hist = px.line(hist_df, x=hist_df.index, y='Close', title=f'{company_nm} Price History vs. DCF'); fig_hist.update_layout(xaxis_title='Date', yaxis_title=f'Price ({currency_sym})', height=400, legend_title_text='Legend'); fig_hist.update_xaxes(rangeslider_visible=False)
        if valid_intrinsic: fig_hist.add_hline(y=intrinsic_val_num, line_dash="dash", line_color="#e74c3c", opacity=0.8, annotation_text=f"DCF Intrinsic: {format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian)}", annotation_position="bottom right", annotation_font_color="#e74c3c")
        if valid_current_pr: fig_hist.add_hline(y=current_pr_num, line_dash="dot", line_color="#3498db", opacity=0.8, annotation_text=f"Current: {format_currency(current_pr_num, currency_sym, is_indian=is_indian)}", annotation_position="top left", annotation_font_color="#3498db")
        fig_hist.show()
    else: display(Markdown("*Historical price data unavailable.*"))

    # --- DCF EV Breakdown Pie Chart ---
    # Ensure fcff_res is a dictionary before accessing
    if isinstance(fcff_res, dict):
        pv_forecast_sum = fcff_res.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_res.get('pv_terminal_value', np.nan)
        valid_pvs = np.isfinite([pv_forecast_sum, pv_tv]); total_ev = (pv_forecast_sum + pv_tv) if all(valid_pvs) else np.nan
        if np.isfinite(total_ev) and total_ev != 0 :
            labels_pie = []; values_pie = []; pull_pie = []; colors = ['#4e79a7','#f28e2b']
            if abs(pv_forecast_sum) > 1e-6: labels_pie.append(f'PV({FORECAST_YEARS}yr FCFFs)'); values_pie.append(pv_forecast_sum); pull_pie.append(0)
            if abs(pv_tv) > 1e-6: labels_pie.append('PV(TV)'); values_pie.append(pv_tv); pull_pie.append(0.05 if len(labels_pie) > 1 else 0)
            if labels_pie:
                hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                if is_indian: hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                fig_dcf_pie = go.Figure(data=[go.Pie(labels=labels_pie, values=values_pie, hole=.4, pull=pull_pie, marker_colors=colors[:len(labels_pie)], hovertemplate=hover_format, textinfo='percent+label', insidetextorientation='radial', sort=False)])
                fig_dcf_pie.update_layout( title_text=f"DCF EV Breakdown (Total EV: {format_currency(total_ev, currency_sym, is_indian=is_indian)})", title_x=0.5, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), height=350, margin=dict(t=60, b=50, l=20, r=20) )
                fig_dcf_pie.show()
            else: display(Markdown(f"*DCF components zero/negligible. Cannot generate EV Breakdown chart.*"))
        else: display(Markdown(f"*EV calculation failed ({format_currency(total_ev, currency_sym, is_indian=is_indian)}). Cannot generate EV Breakdown chart.*"))
    else: display(Markdown(f"*FCFF results invalid. Cannot generate EV Breakdown chart.*"))

    # --- FCFF Projection Bar Chart ---
    # Ensure fcff_res is a dictionary before accessing
    if isinstance(fcff_res, dict) and fcff_res.get('projected_fcffs'):
        years = [f'Y{i+1}' for i in range(len(fcff_res['projected_fcffs']))]; pv_list_clean = [p if np.isfinite(p) else 0 for p in fcff_res.get('pv_fcffs', [])]; fcff_list_clean = [f if np.isfinite(f) else 0 for f in fcff_res.get('projected_fcffs', [])]
        if len(pv_list_clean) == len(years) and len(fcff_list_clean) == len(years):
            df_fcff_plot = pd.DataFrame({'Year': years, 'Forecasted FCFF': fcff_list_clean, 'PV of FCFF': pv_list_clean}); df_fcff_plot['Forecasted FCFF'] = pd.to_numeric(df_fcff_plot['Forecasted FCFF']); df_fcff_plot['PV of FCFF'] = pd.to_numeric(df_fcff_plot['PV of FCFF'])
            df_fcff_melt = df_fcff_plot.melt(id_vars='Year', var_name='Cash Flow Type', value_name='Value')
            fig_fcff_bar = px.bar(df_fcff_melt, x='Year', y='Value', color='Cash Flow Type', title='FCFF Forecast & Present Values', labels={'Value':f'Value ({currency_sym})'}, barmode='group', height=350, color_discrete_map={'Forecasted FCFF':'#5dade2', 'PV of FCFF':'#af7ac5'})
            hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            if is_indian: hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            fig_fcff_bar.update_traces(hovertemplate=hover_bar_format); fig_fcff_bar.update_layout(margin=dict(t=60, b=30, l=50, r=30)); fig_fcff_bar.show()
        else: display(Markdown("*FCFF projection data mismatch/invalid. Cannot plot.*"))
    else: display(Markdown("*FCFF projection data unavailable or invalid.*"))

    # --- Valuation Conclusion ---
    display(Markdown("### Valuation Conclusion"));
    # Ensure final_val_dict is a dictionary before accessing
    if isinstance(final_val_dict, dict):
        intrinsic_val = final_val_dict.get('intrinsic_value_per_share', np.nan)
        current_pr_num = current_pr if valid_current_pr else np.nan; conclusion = ""
        if final_val_dict.get('error'): conclusion = f"<span style='color:red'>❌ Valuation Error: {final_val_dict.get('error', '?')}</span>"
        elif not np.isfinite(intrinsic_val) or intrinsic_val <= 0: conclusion = f"⚠️ Valuation resulted in invalid/non-positive Intrinsic Value ({format_currency(intrinsic_val, currency_sym, is_indian=is_indian)})."
        elif not np.isfinite(current_pr_num): conclusion = f"✅ Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}. Cannot compare (Current Price N/A)."
        else:
             diff_pct = (intrinsic_val / current_pr_num) - 1
             current_fmt = format_currency(current_pr_num, currency_sym, is_indian=is_indian)
             intrinsic_fmt = format_currency(intrinsic_val, currency_sym, is_indian=is_indian)
             if abs(diff_pct) < 0.10: conclusion = f"<span style='color:grey'>~ Fairly Valued:</span> Price ({current_fmt}) vs DCF ({intrinsic_fmt}). Diff: {diff_pct:+.1%}"
             elif diff_pct < 0: overval_pct = (current_pr_num / intrinsic_val) - 1; conclusion = f"<span style='color:#e74c3c'>🔻 Potentially Overvalued:</span> Price ({current_fmt}) is **{overval_pct:+.1%}** above DCF ({intrinsic_fmt})."
             else: conclusion = f"<span style='color:#2ecc71'>✅ Potentially Undervalued:</span> Price ({current_fmt}) suggests **{diff_pct:+.1%}** upside to DCF ({intrinsic_fmt})."
    else:
        conclusion = f"<span style='color:red'>❌ Valuation Error: Final valuation data structure invalid.</span>"

    display(Markdown(f"> **{conclusion}**"))

# --- Execute Summary Dashboard Generation ---
# Check if results are valid dictionaries before passing them
if isinstance(final_valuation, dict) and isinstance(fcff_results, dict) and isinstance(stock_data.get('info'), dict) and isinstance(input_sources, dict) and not final_valuation.get('error'):
    try:
        display_summary_dashboard(
            company_name, ticker_symbol, current_price, final_valuation.get('intrinsic_value_per_share'),
            discount_rate, fcff_results.get('terminal_growth_rate'), fcff_results.get('initial_growth_rate'),
            fcff_results.get('growth_source', 'N/A'), final_valuation, stock_data['history'], fcff_results,
            currency_symbol, stock_data['info'], input_sources, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Dashboard Generation: {e}**"))
else:
    display(Markdown("❌ **Skipping Summary Dashboard due to prior errors or invalid data structures.**"))
    # Optionally print the error states for debugging:
    # print("Final Valuation State:", final_valuation)
    # print("FCFF Results State:", fcff_results)


# %% [markdown]
# ## 10. Assumptions & Disclaimer (Reflects Dynamic/User Inputs & Fallbacks - Growth Logic Updated)

# %%
# Dynamically generate assumption strings based on final values
# Add checks to ensure fcff_results and wacc_components are dictionaries before accessing
rf_source_disp = input_sources.get('Rf', '?')
rm_source_disp = input_sources.get('Rm','?')
beta_source_disp = input_sources.get('Beta','?')
c_spread_source_disp = input_sources.get('Credit Spread','N/A')
bench_source_disp = input_sources.get('Benchmarks','?')

# Safely get values from dictionaries, providing defaults if dicts don't exist or keys missing
growth_source_disp = fcff_results.get('growth_source', 'N/A') if isinstance(fcff_results, dict) else 'Error/Not Run'
tax_rate_source_disp = wacc_components.get('tax_rate_source', 'N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
equity_source_disp = wacc_components.get('equity_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
weights_source_disp = wacc_components.get('weights_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
terminal_growth_final = fcff_results.get('terminal_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan
initial_growth_final = fcff_results.get('initial_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan


# %%
# ----> UPDATED TEXT TO REFLECT NEW GROWTH LOGIC (V13) <----
display(Markdown(f"""
**Key Assumptions & Input Sources:**
*   **Financial Data:** Relies on data provided by Yahoo Finance (`yfinance`). Accuracy subject to provider. Missing data uses fallbacks or causes errors.
*   **Risk-Free Rate (Rf):** Attempted dynamic fetch, **confirmed/overridden by user.** (`Source: {rf_source_disp}` - Default suggestion: `{DEFAULT_RF_RATE_SUGGESTION:.1f}%`)
*   **Expected Market Return (Rm):** Based on **user input.** (`Source: {rm_source_disp}`)
*   **Market Risk Premium (MRP):** Derived from user `Rm` and final `Rf`.
*   **Beta (β):** Dynamically fetched if valid, otherwise **required user input.** (`Source: {beta_source_disp}`)
*   **Cost of Debt (Kd):** Implied if possible & within bounds (`{format_percent(MIN_COST_OF_DEBT)}` - `{format_percent(MAX_COST_OF_DEBT)}`). Otherwise, defaults to `Rf + Credit Spread`.
*   **Credit Spread:** Used only if default Kd triggered. Based on **user input** (default suggestion: `{format_percent(DEFAULT_CREDIT_SPREAD)}`). (`Source: {c_spread_source_disp}`)
*   **Growth Rates:**
    *   **Initial Growth:** `{format_percent(initial_growth_final)}` used for Year 1, determined via: (`Source: {growth_source_disp}`)
        1.  Fetch last 3 years (4 data points) of Revenue, FCFF, and Stock Price (daily).
        2.  **Validate Data:** Check if T-0 (end) is finite and T-3 (start) is positive and finite for each metric.
        3.  Calculate **3-Year CAGR** *only* for metrics with valid start/end data (using `calculate_cagr_modified`). Stock CAGR uses daily price history, preferring 'Adj Close', falling back to 'Close'.
        4.  Calculate the **Geometric Mean** (`geometric_mean_modified`) of the successfully calculated CAGRs from step 3.
        5.  If GM calculation fails or *no valid CAGRs were computed* in step 3, use **Default (`{format_percent(DEFAULT_GROWTH_RATE)}`)**.
        6.  Apply **Bounds:** The result from step 4 or 5 is capped between **`{format_percent(MIN_INITIAL_GROWTH_FLOOR)}` (Floor)** and **`{format_percent(MAX_INITIAL_GROWTH_CEILING)}` (Ceiling)**.
    *   **Terminal Growth (g):** `{format_percent(terminal_growth_final)}` = min(GDP cap=`{format_percent(TERMINAL_GROWTH_RATE_CAP)}`, WACC*`0.67`, Final Initial Growth, Rf*`0.95`), floored at `{format_percent(MIN_GROWTH)}`, ensured `< WACC`.
*   **Tax Rate:** Effective rate `{tax_rate_source_disp}` (bounded `{format_percent(MIN_TAX_RATE)}` - `{format_percent(MAX_TAX_RATE)}`) or default (`{format_percent(DEFAULT_TAX_RATE)}`).
*   **Capital Structure:** Assumed stable using `{equity_source_disp}` and calculated debt. (`Weights Source: {weights_source_disp}`)
*   **Industry Benchmarks:** Uses **user input** or **generic fallbacks** if input left blank. (`Source: {bench_source_disp}`)
*   **Currency Formatting:** Uses ₹/Lakh/Crore for `.NS` tickers, standard K/M/B/T otherwise.

**Limitations:**
*   **Model Sensitivity:** DCF highly sensitive to inputs. **Sensitivity analysis (not included) crucial.**
*   **Input Quality:** Validity depends on user inputs and appropriateness of fallbacks.
*   **Historical Data Availability:** Financial statement data (needed for Rev/FCFF CAGR) is often limited to ~4 years via `yfinance`. Accuracy of 3-year CAGR depends heavily on these few points. Stock price history is usually more extensive but relies on finding suitable start/end dates. CAGR calculation is skipped if start/end data is invalid or start is non-positive.
*   **CAGR/GM Modifications:** The modified CAGR handles non-positive end values (relative to positive start) but requires a positive start value. The modified GM handles negative rates but falls back to arithmetic mean if rates are <= -100% or other errors occur. GM is less meaningful with highly disparate or negative rates.
*   **Dynamic Data Reliability:** Depends on `yfinance` accuracy/availability. Rf proxies need verification. Stock history column names ('Adj Close', 'Close') assumed.
*   **Simplifications:** Linear decay, constant rates/structure. Ignores qualitative factors, macro shocks, sentiment.
*   **Fallback Accuracy:** Default growth rate (6.5%) and generic benchmarks may not be representative.
*   **Data Completeness:** Performance degrades with missing data. *Cross-verify key data.*

**Disclaimer:**
*For **educational and illustrative purposes only**. **NOT investment advice**. Models have limitations. Users MUST conduct own due diligence and consult a qualified financial advisor before investing. Creators NOT liable for actions based on this report.*
"""))


# %%\
````

````python
# %% [markdown]
# # Dynamic Stock Financial Analysis & DCF Valuation
# **(Interactive Report - Dynamic/User Input Driven Assumptions - V14 - Resilient Growth & Projection)**
#
# **Analyst:** Simulated AI Research Analyst
#
# **Objective:** To perform a comprehensive financial analysis and estimate the intrinsic value per share of a publicly traded company using the Discounted Cash Flow (DCF) methodology. Prioritizes dynamic fetching, falls back to user input, and uses configured defaults as a last resort. Includes fallback benchmarks for ratios and robust error handling/display. Adapts currency formatting for Indian stocks. **Initial growth uses a Geometric Mean of valid 3yr Rev/FCFF/Stock CAGRs (modified formulas for negatives), bounded & defaulted. Handles cases where FCFF T-0 is invalid for projection but still attempts growth calc.**
#
# **Methodology:**
# 1.  **User Input & Dynamic Fetching:** Get ticker. Attempt dynamic Rf, require user confirmation/input for Rf, Rm, Beta (if needed), Credit Spread (if needed), and Ratio Benchmarks (with fallbacks). Detects Indian stocks for currency formatting.
# 2.  **Data Acquisition:** Use `yfinance` for statements, info, market data. Fetch extended price history (4y, daily interval). Inspect fetched columns.
# 3.  **WACC Calculation:** Determine WACC using fetched/user inputs via CAPM. Handle errors.
# 4.  **FCFF Calculation & Projection:**
#     *   Calculate historical FCFF (T-0 to T-3). Log if T-0 FCFF is invalid.
#     *   **Estimate initial growth:**
#         *   Fetch last 3 years (4 data points) of Revenue, FCFF, and Stock Price (daily).
#         *   Check validity of start (T-3) and end (T-0) points for each metric.
#         *   Calculate 3-year CAGR *only* for metrics with valid data (using modified CAGR). Stock CAGR uses 'Adj Close' or 'Close'.
#         *   Calculate the Geometric Mean of the valid calculated CAGRs (using modified GM).
#         *   Fall back to default (6.5%) if GM fails or no valid CAGRs exist.
#         *   Bound the result between 1.1% and 15%.
#     *   **Project FCFF (Conditional):** Check if `fcff_current` (T-0) is valid *after* growth rate estimation. If invalid, **stop projection** and set error, otherwise proceed.
#     *   Project FCFF forward using linear decay to terminal growth. Handle errors.
# 5.  **Terminal Value:** Estimate value beyond the explicit forecast period (only if projection occurred).
# 6.  **Discounting & Valuation:** Discount cash flows and terminal value. Handle errors. Initialize result dicts. Skips if projection failed.
# 7.  **Ratio Analysis:** Calculate ratios and compare to user-input or fallback default benchmarks. Handles non-finite values. Initialize result dicts. Skips if valuation failed.
# 8.  **Interactive Visualization:** Present findings using Plotly, including dynamic gauge ranges, robust display logic, and ₹/Lakh/Crore formatting for Indian stocks.

# %% [markdown]
# ## 1. Setup and Libraries

# %%
# Standard Libraries
import warnings
import sys
import numpy as np
import pandas as pd
import datetime
import math # For isnan, isinf checks

# Data Fetching
import yfinance as yf

# Visualization (Using Plotly)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

# Display Utilities
from IPython.display import display, Markdown, HTML

# Settings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pio.templates.default = "plotly_white" # Cleaner Plotly template

# %% [markdown]
# ## 2. Configuration and Helper Functions (Growth Helpers Updated)

# %%
# --- Model Configuration ---
DEFAULT_TICKER = "WIPRO.NS" # Default if user input fails
FORECAST_YEARS = 5
# Terminal Growth Rate CAP (Also acts as a default ceiling)
TERMINAL_GROWTH_RATE_CAP = 0.05 # Default/Cap set to 5%
# ----> DEFAULT INITIAL GROWTH (Used only if GM of CAGRs calc fails) <----
DEFAULT_GROWTH_RATE = 0.065  # Default set to 6.5%
# Default Risk-Free Rate suggestion (used if dynamic fetch fails)
DEFAULT_RF_RATE_SUGGESTION = 6.0 # Default set to 6.0% (as percent)

DEFAULT_CURRENCY_SYMBOL = "₹" # Changed default to Rupee, will be overridden by yfinance if possible
DEFAULT_TAX_RATE = 0.25     # General fallback tax rate if calculation fails
DEFAULT_CREDIT_SPREAD = 0.03 # Fallback spread over Rf for Cost of Debt
DEFAULT_MARKET_RISK_PREMIUM = 0.08 # Used only to suggest Rm to user

# --- Generic Fallback Benchmark Ratios (Used ONLY if User leaves input blank) ---
DEFAULT_BENCHMARK_RATIOS = {
    'P/E': 20.0,
    'P/B': 2.5,
    'EV/EBITDA': 12.0,
    'ROE': 15.0, # As percent value (e.g., 15 for 15%)
    'ROA': 8.0,  # As percent value
    'ROCE': 12.0,# As percent value
    'Asset Turnover': 0.8,
    'Current Ratio': 1.5,
    'Quick Ratio': 1.0
}

# --- Model Stability Bounds --- (Not primary financial assumptions, but guardrails)
MIN_WACC = 0.06
MAX_WACC = 0.25
MIN_TAX_RATE = 0.10
MAX_TAX_RATE = 0.40
MIN_GROWTH = 0.00 # Minimum growth floor (for terminal growth and projection steps)
# ----> Bounds for CALCULATED INITIAL GROWTH RATE <----
MIN_INITIAL_GROWTH_FLOOR = 0.011 # 1.1%
MAX_INITIAL_GROWTH_CEILING = 0.15  # 15.0%

MIN_COST_OF_DEBT = 0.03
MAX_COST_OF_DEBT = 0.15

# --- Ratio Keys Needed for Benchmarking ---
REQUIRED_BENCHMARK_RATIOS = list(DEFAULT_BENCHMARK_RATIOS.keys()) # Use keys from default dict

# --- Helper Functions ---
def print_keys(statement, name):
    """Prints available keys in a yfinance statement for debugging"""
    print(f"\n--- Available Keys in {name} ---")
    if isinstance(statement, pd.DataFrame):
        valid_indices = [idx for idx in statement.index if idx is not None]
        print(", ".join(map(str, valid_indices)))
    elif isinstance(statement, dict):
        print(", ".join(statement.keys()))
    print("------------------------------------\n")

def safe_get(data, key_list, statement_name="Statement", default=0.0):
    """
    Safely retrieves a numeric value from yfinance data (Series or Dict).
    Tries multiple keys, returns default if not found/invalid. Handles NaN/inf better.
    """
    if not isinstance(key_list, list):
        key_list = [key_list]

    value = None
    for key in key_list:
      current_value = None
      raw_val = None
      if isinstance(data, pd.Series):
          if key in data.index:
              try:
                  raw_val = data.loc[key]
                  # Check if raw_val is already numeric and finite before converting
                  if isinstance(raw_val, (int, float)) and np.isfinite(raw_val):
                      current_value = float(raw_val)
                  # Handle Series case (sometimes yfinance returns single value Series)
                  elif isinstance(raw_val, pd.Series):
                       if not raw_val.empty and pd.notna(raw_val.iloc[0]) and np.isfinite(raw_val.iloc[0]):
                            current_value = float(raw_val.iloc[0])
                  # Handle string conversion only if necessary and check finiteness
                  elif pd.notna(raw_val):
                      try_val = float(raw_val)
                      if np.isfinite(try_val):
                          current_value = try_val
              except (KeyError, TypeError, IndexError, ValueError):
                   pass # Ignore errors for this specific key/conversion
      elif isinstance(data, dict):
          raw_val = data.get(key)
          if raw_val is not None:
              try:
                  if isinstance(raw_val, str) and raw_val.lower() in ['n/a', 'none', '']:
                      pass # Treat as invalid
                  else:
                      temp_val = float(raw_val)
                      if np.isfinite(temp_val): # Check for NaN/inf after conversion
                           current_value = temp_val
              except (TypeError, ValueError):
                   pass

      if current_value is not None:
            value = current_value
            break # Found a valid value

    if value is None:
        value = default # Use the provided default if no valid value found
    return value

def format_currency(value, symbol="₹", precision=2, is_indian=False):
    """
    Formats number as currency string, handles None/NaN/inf.
    Uses Lakh/Crore for Indian stocks if is_indian=True.
    """
    if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
    if abs(value) > 1e18: return "Inf" # Handle extremely large numbers as Inf

    sign = "-" if value < 0 else ""
    abs_val = abs(value)

    if is_indian:
        # Use ₹ symbol regardless of fetched symbol for consistency
        symbol = "₹"
        if abs_val >= 1e7:  # 1 Crore (10 million)
            formatted_val = f"{abs_val / 1e7:.{precision}f} Cr"
        elif abs_val >= 1e5:  # 1 Lakh (100 thousand)
            formatted_val = f"{abs_val / 1e5:.{precision}f} Lakh"
        else:
            formatted_val = f"{abs_val:,.{precision}f}" # Add comma separators for smaller Rupee values
        return f"{sign}{symbol} {formatted_val}" # Add space after symbol
    else:
        # Standard international formatting (K, M, B, T)
        if abs_val >= 1e12: formatted_val = f"{abs_val / 1e12:.{precision}f}T"
        elif abs_val >= 1e9: formatted_val = f"{abs_val / 1e9:.{precision}f}B"
        elif abs_val >= 1e6: formatted_val = f"{abs_val / 1e6:.{precision}f}M"
        elif abs_val >= 1e3: formatted_val = f"{abs_val / 1e3:.{precision}f}K"
        else: formatted_val = f"{abs_val:.{precision}f}"
        return f"{sign}{symbol}{formatted_val}" # Standard symbol prefix

def format_percent(value, precision=1):
     """Formats number as percentage string, handles None/NaN/inf."""
     if value is None or pd.isna(value) or not math.isfinite(value): # Use math.isfinite
        return "N/A"
     if abs(value) > 1e5: return "Inf" # Handle extreme % values

     return f"{value * 100:.{precision}f}%"

def get_validated_float_input(prompt_message, default_value=None, allow_blank=False, min_val=None, max_val=None):
    """Gets and validates user input for a float value, handles defaults, blanks, and bounds."""
    while True:
        prompt_with_default = prompt_message
        if default_value is not None:
            prompt_with_default += f" [Default: {default_value}]: "
        elif allow_blank:
             prompt_with_default += f" [Leave blank to skip/use fallback]: "
        else:
             prompt_with_default += ": "

        try:
            user_input_str = input(prompt_with_default).strip()

            if not user_input_str:
                if allow_blank: return None
                if default_value is not None:
                    try: return float(default_value)
                    except ValueError: raise ValueError(f"Internal Error: Invalid default value '{default_value}'.")
                raise ValueError("Input cannot be empty.")

            if user_input_str.endswith('%'): # Allow user to add % sign optionally
                user_input_str = user_input_str[:-1]

            value = float(user_input_str)

            if min_val is not None and value < min_val:
                 raise ValueError(f"Value must be >= {min_val}.")
            if max_val is not None and value > max_val:
                 raise ValueError(f"Value must be <= {max_val}.")

            return value
        except ValueError as e:
            print(f"❌ Invalid input: {e}. Please enter a valid number.")

def try_get_dynamic_rf(rf_ticker_symbol="^TNX"):
    """Attempts to fetch a dynamic risk-free rate (e.g., US 10Y Treasury)."""
    print(f"Attempting to fetch dynamic Risk-Free Rate ({rf_ticker_symbol})...")
    try:
        rf_ticker = yf.Ticker(rf_ticker_symbol)
        hist = rf_ticker.history(period="5d")
        if not hist.empty and 'Close' in hist.columns:
            last_close = hist['Close'].iloc[-1]
            if pd.notna(last_close):
                dynamic_rf = last_close / 100.0 # Assume Yahoo gives yield as %
                if 0.001 < dynamic_rf < 0.20: # Check bounds 0.1% to 20%
                    print(f"✅ Dynamically fetched Rf ({rf_ticker_symbol}): {format_percent(dynamic_rf)}")
                    return dynamic_rf
                else:
                     print(f"⚠️ Fetched Rf value ({format_percent(dynamic_rf)}) out of typical range. Ignoring.")
                     return None
            else:
                 print(f"⚠️ Fetched Rf data was invalid (NaN) for {rf_ticker_symbol}.")
                 return None
        else:
             print(f"⚠️ Could not fetch recent data for {rf_ticker_symbol}.")
             return None
    except Exception as e:
        print(f"❌ Error fetching dynamic Rf ({rf_ticker_symbol}): {e}")
        return None

# --- Growth Calculation Helpers (Updated) ---
# calculate_average_growth kept for reference
def calculate_average_growth(values_list):
    """Calculates average simple growth rate (NOT CAGR). Kept for reference."""
    growth_rates = []
    if len(values_list) < 2: return np.nan
    for i in range(len(values_list) - 1):
        val_t = values_list[i]; val_t_minus_1 = values_list[i+1]
        if pd.notna(val_t) and pd.notna(val_t_minus_1) and math.isfinite(val_t) and math.isfinite(val_t_minus_1):
            if abs(val_t_minus_1) > 1e-9:
                g = (val_t / val_t_minus_1) - 1
                if -10.0 < g < 10.0: growth_rates.append(g)
    return np.mean(growth_rates) if growth_rates else np.nan

def calculate_cagr_modified(start_val, end_val, num_years):
    """
    Calculates Compound Annual Growth Rate (CAGR).
    MODIFIED to handle cases where end_val is negative or zero, assuming start_val is positive.
    Returns NaN if start_val is non-positive, num_years <= 0, or values are invalid.
    Also returns NaN if result is extremely high/low (often indicates data issues).
    """
    if pd.isna(start_val) or pd.isna(end_val) or not math.isfinite(start_val) or not math.isfinite(end_val) or num_years <= 0:
        return np.nan
    # CRITICAL: Start value must be positive for any meaningful CAGR interpretation.
    if start_val <= 1e-9:
        return np.nan

    cagr = np.nan # Initialize cagr as nan
    try:
        # Handle non-positive end value
        if end_val <= 1e-9:
            end_val_abs = max(abs(end_val), 1e-9)
            ratio = end_val_abs / start_val
            if ratio > 0:
                 # Calculate magnitude and make it negative
                 cagr_magnitude = ratio ** (1 / num_years) - 1
                 cagr = -cagr_magnitude
            # else: cagr remains NaN
        else:
            # Standard CAGR calculation if both start and end are positive
            ratio = end_val / start_val
            if ratio > 0:
                 cagr = ratio ** (1 / num_years) - 1
            # else: cagr remains NaN

    except (ValueError, OverflowError, ZeroDivisionError) as e:
        cagr = np.nan # Ensure cagr is NaN on error

    # Final check for extremely large/small results
    if pd.notna(cagr) and not (-0.9999 < cagr < 10.0): # Reject >1000% or <-99.99% growth
        cagr = np.nan

    return cagr


def geometric_mean_modified(rates):
    """
    Calculates the geometric mean of a list of rates (e.g., CAGRs).
    MODIFIED using an offset method to handle negative rates (down to nearly -100%).
    Falls back to arithmetic mean if any rate is <= -1.0 (-100%) or calculation fails.
    Returns NaN if no valid rates are provided.
    """
    # Filter out non-finite values first
    valid_rates = [r for r in rates if pd.notna(r) and math.isfinite(r)]
    if not valid_rates:
        return np.nan
    if len(valid_rates) == 1:
        return valid_rates[0] # GM of one item is the item itself

    # Offset method: Add 1 to each rate. This factor MUST be > 0.
    offset = 1.0
    factors_offset = [r + offset for r in valid_rates]

    # Check if any rate was <= -1.0 (which makes the factor <= 0)
    if any(f <= 1e-9 for f in factors_offset): # Use a small threshold
         print("⚠️ Geometric mean calculation failed due to rate <= -100%. Falling back to arithmetic mean.")
         return np.mean(valid_rates) # Fallback uses only the valid rates

    n = len(valid_rates)
    product = 1.0
    try:
        for factor in factors_offset:
             product *= factor
        # Handle potential complex numbers if product is negative and n is even
        if product < 0 and n % 2 == 0:
             print("⚠️ Geometric mean calculation failed (negative base for even root). Falling back to arithmetic mean.")
             return np.mean(valid_rates)

        # Calculate GM of offset rates (sign handling for odd roots of negative numbers)
        gm_offset = np.sign(product) * (np.abs(product) ** (1/n))

        # Subtract the offset
        gm = gm_offset - offset

        # Add a check for extreme GM results
        if pd.notna(gm) and not (-0.9999 < gm < 10.0):
            print(f"⚠️ Geometric mean result ({format_percent(gm)}) out of reasonable bounds. Falling back to arithmetic mean.")
            return np.mean(valid_rates)

        return gm
    except (ValueError, OverflowError, ZeroDivisionError, TypeError) as e: # Added TypeError
        print(f"⚠️ Geometric mean calculation encountered numeric error ({e}). Falling back to arithmetic mean.")
        return np.mean(valid_rates)

# --- End Growth Calculation Helpers ---


# %% [markdown]
# ## 3. User Input & Dynamic Fetching: Ticker, Rf, Rm, Benchmarks, Beta, Spread

# %%
input_sources = {}

# --- Get Ticker Symbol ---
user_ticker = input(f"Enter stock ticker (e.g., MSFT, AAPL, RELIANCE.NS) [Default: {DEFAULT_TICKER}]: ")
if not user_ticker: user_ticker = DEFAULT_TICKER
ticker_symbol = user_ticker.strip().upper()
display(Markdown(f"**Analyzing Ticker:** `{ticker_symbol}`"))

# ----> DETECT IF INDIAN STOCK for currency formatting <----
is_indian_stock = ticker_symbol.endswith(".NS")
if is_indian_stock:
    print("Info: Detected Indian stock (.NS). Using ₹/Lakh/Crore formatting.")

# --- Determine RF Ticker based on Stock Ticker ---
rf_proxy_ticker = "^TNX" # Default US 10Y
if is_indian_stock: rf_proxy_ticker = "^NS10"; print(f"Info: Using Indian Rf proxy: {rf_proxy_ticker}")
elif ticker_symbol.endswith(".L"): rf_proxy_ticker = "^FTLC"; print(f"Info: Using UK Rf proxy: {rf_proxy_ticker}")
# Add more country/exchange specific Rf proxies if needed

# --- Get Risk-Free Rate (Rf) ---
print("\n--- CAPM Inputs ---")
dynamic_rf_rate = try_get_dynamic_rf(rf_proxy_ticker)
rf_prompt = f"Enter Risk-Free Rate (Rf) as percentage (e.g., '4.5' for 4.5%). Fetched {rf_proxy_ticker}?"
suggested_rf_val = dynamic_rf_rate * 100 if dynamic_rf_rate is not None else DEFAULT_RF_RATE_SUGGESTION

risk_free_rate_input_pct = get_validated_float_input(
    rf_prompt, default_value=f"{suggested_rf_val:.2f}", allow_blank=False, min_val=0.0, max_val=20.0
)
risk_free_rate_input = risk_free_rate_input_pct / 100.0

input_sources['Rf'] = 'Dynamic Fetch (User Confirmed/Adjusted)' if dynamic_rf_rate is not None else f'User Input (Default: {DEFAULT_RF_RATE_SUGGESTION:.1f}%)'
display(Markdown(f"**Using Risk-Free Rate (Rf):** {format_percent(risk_free_rate_input)} (`Source: {input_sources['Rf']}`)"))

# --- Get Expected Market Return (Rm) ---
suggested_mrp = DEFAULT_MARKET_RISK_PREMIUM
suggested_rm = risk_free_rate_input + suggested_mrp
rm_prompt = f"Enter Expected Market Return (Rm) as percentage (e.g., '10'). Suggestion based on Rf + {format_percent(suggested_mrp)} MRP is {format_percent(suggested_rm)}"

while True:
    market_return_input_pct = get_validated_float_input(
        rm_prompt, default_value=f"{suggested_rm * 100:.1f}", allow_blank=False, min_val=-10.0, max_val=35.0
    )
    market_return_input = market_return_input_pct / 100.0
    if market_return_input >= risk_free_rate_input: break
    else: print(f"❌ Error: Rm ({format_percent(market_return_input)}) must be >= Rf ({format_percent(risk_free_rate_input)}). Re-enter.")

market_risk_premium_input = market_return_input - risk_free_rate_input
input_sources['Rm'] = 'User Input'
input_sources['MRP'] = 'Derived'
display(Markdown(f"**Using Expected Market Return (Rm):** {format_percent(market_return_input)} (`Source: {input_sources['Rm']}`)"))
display(Markdown(f"**↳ Implied Market Risk Premium (MRP):** {format_percent(market_risk_premium_input)} (`Source: {input_sources['MRP']}`)"))

# --- Beta (Placeholder) ---
beta_input = None; input_sources['Beta'] = 'Pending'
# --- Credit Spread (Placeholder) ---
user_credit_spread = None; input_sources['Credit Spread'] = 'Pending'

# --- Get Industry Benchmark Ratios (with fallback defaults) ---
print("\n--- Industry Ratio Benchmarks ---")
display(Markdown("Enter benchmarks or leave blank for generic fallbacks. *(Find industry averages online)*"))
user_benchmarks = {}; benchmark_sources = {}
print("Provide benchmarks (e.g., ROE: enter '15' for 15%; P/E: enter '20'):")
for ratio_key in REQUIRED_BENCHMARK_RATIOS:
    default_val = DEFAULT_BENCHMARK_RATIOS.get(ratio_key, None)
    prompt = f"  Enter benchmark for {ratio_key}"
    is_percent_type = ratio_key in ['ROE', 'ROA', 'ROCE']
    min_b, max_b = (-50, 100) if is_percent_type else (-100, 500)

    bench_val_input = get_validated_float_input(
        prompt, default_value=f"{default_val:.2f}" if default_val is not None else None,
        allow_blank=True, min_val=min_b, max_val=max_b
    )

    if bench_val_input is not None:
        user_benchmarks[ratio_key] = bench_val_input
        benchmark_sources[ratio_key] = 'User Input'
        display_val = f"{bench_val_input:.1f}%" if is_percent_type else f"{bench_val_input:.2f}"
        display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: User`)"))
    elif default_val is not None:
         user_benchmarks[ratio_key] = default_val # Store the default value if user skipped
         benchmark_sources[ratio_key] = 'Fallback Default'
         display_val = f"{default_val:.1f}%" if is_percent_type else f"{default_val:.2f}"
         display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: Default`)"))
    else:
        benchmark_sources[ratio_key] = 'Not Available'
        display(Markdown(f"  - **{ratio_key} Benchmark:** Not Provided / No Default"))
input_sources['Benchmarks'] = 'User Input or Default'

# %% [markdown]
# ## 4. Data Acquisition (Fetching Extended History)

# %%
stock_data = {}
try:
    stock = yf.Ticker(ticker_symbol)
    info = stock.info
    if not info or not info.get('regularMarketPrice'):
        print(f"Warning: Limited info for {ticker_symbol}. Trying history...")
        hist_check = stock.history(period="5d");
        if hist_check.empty: raise ValueError(f"Ticker '{ticker_symbol}' invalid/no data.")
        else: print(f"Warning: Proceeding with limited data."); info = {} if not isinstance(info, dict) else info
    stock_data['info'] = info

    # Standard statement fetching (usually 4 years)
    stock_data['financials'] = stock.financials
    stock_data['balance_sheet'] = stock.balance_sheet
    stock_data['cashflow'] = stock.cashflow
    if stock_data['financials'].empty or stock_data['balance_sheet'].empty or stock_data['cashflow'].empty:
       print(f"⚠️ Warning: Empty statements for {ticker_symbol}. DCF unreliable.")

    # Fetch recent history for charts/current price fallback
    stock_data['history'] = stock.history(period="2y")
    # ----> Fetch extended history (need at least ~3 years for 3yr CAGR) <----
    # Fetch 4 years daily data to maximize chance of finding start date ~3 years ago
    stock_data['history_long'] = stock.history(period="4y", interval="1d") # Explicitly daily
    # ----> INSPECT COLUMNS <----
    if not stock_data['history_long'].empty:
        print(f"Info: Fetched long history (4y daily). Columns available: {list(stock_data['history_long'].columns)}")
    else:
         print(f"⚠️ Warning: Could not fetch long history (4y daily) for {ticker_symbol}. Stock CAGR calc may fail.")
    if stock_data['history_long'].empty or len(stock_data['history_long']) < 2: # Need at least 2 points
        print(f"⚠️ Warning: Fetched long history has < 2 data points. Stock CAGR calc will fail.")


    company_name = info.get("longName", info.get("shortName", ticker_symbol))
    # Determine currency symbol, prioritizing Rupee for Indian stocks
    currency_code = info.get("currency", DEFAULT_CURRENCY_SYMBOL)
    if not isinstance(currency_code, str) or len(currency_code)>5 : currency_code=DEFAULT_CURRENCY_SYMBOL
    currency_symbol = "₹" if is_indian_stock else info.get("symbol", currency_code) # Prefer symbol if available, else code, force Rupee for Indian

    current_price = safe_get(info, ['currentPrice', 'regularMarketPrice', 'previousClose', 'lastPrice'], 'Info', default=np.nan)
    price_source = "yfinance Info"
    if not np.isfinite(current_price) and not stock_data['history'].empty:
        last_close_hist = stock_data['history']['Close'].iloc[-1]
        if pd.notna(last_close_hist): current_price = last_close_hist; price_source = "yfinance History"
    if not np.isfinite(current_price): print(f"⚠️ Warning: Could not determine current price."); current_price = np.nan

    shares_outstanding = safe_get(info, ['sharesOutstanding'], 'Info', default=0)

    beta_input = safe_get(info, ['beta'], 'Info', default=None)
    if beta_input is not None and np.isfinite(beta_input):
        input_sources['Beta'] = 'Dynamic Fetch'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))
    else:
         print(f"❌ ERROR: Beta not found dynamically or invalid.")
         beta_input = get_validated_float_input(f"Enter Beta for {ticker_symbol}:", allow_blank=False, min_val=-1.0, max_val=4.0)
         info['beta'] = beta_input; input_sources['Beta'] = 'User Input'
         display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))

    market_cap = safe_get(info, ['marketCap'], 'Info', default=0)
    mcap_source = "yfinance Info"
    if market_cap <= 0 and shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Calculated"; print(f"Info: Market Cap calculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}") # Pass flag
    elif market_cap <= 0: mcap_source = "Unavailable"; print(f"⚠️ Warning: Market Cap unavailable."); market_cap = np.nan

    sector = info.get("sector", "N/A"); industry = info.get("industry", "N/A")
    summary = info.get("longBusinessSummary", "No summary available.")

    display(Markdown(f"**Company:** {company_name}")); display(Markdown(f"**Sector:** {sector} | **Industry:** {industry}"))
    display(Markdown(f"**Currency:** {currency_code} ({currency_symbol})")); # Show both code and effective symbol
    display(Markdown(f"**Current Market Price:** {format_currency(current_price, currency_symbol, is_indian=is_indian_stock)} (`Source: {price_source}`)"))
    display(Markdown(f"**Market Cap:** {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)} (`Source: {mcap_source}`)"))

except Exception as e: print(f"❌ CRITICAL ERROR fetching data: {e}"); sys.exit("Cannot proceed.")

if not isinstance(shares_outstanding, (int, float)) or shares_outstanding <= 0: sys.exit("Essential data missing: Shares Outstanding.")
if beta_input is None or not np.isfinite(beta_input): sys.exit("Essential data missing: Beta.")
if not np.isfinite(market_cap): print(f"⚠️ WARNING: Market Cap unavailable. WACC weights less accurate.")

# %% [markdown]
# ## 5. WACC Calculation

# %%
# WACC Function Definition (No changes here)
def calculate_wacc(stock_info_dict, financials_df, balance_sheet_df,
                   user_rf, user_mrp, user_beta, input_sources_dict,
                   currency_sym, is_indian): # Added currency params
    global user_credit_spread
    display(Markdown("### Calculating WACC..."))
    components = {}
    calculation_log = []
    calculation_log.append(f"Input Sources: Rf='{input_sources_dict.get('Rf')}', MRP='{input_sources_dict.get('MRP')}', Beta='{input_sources_dict.get('Beta')}'")

    beta_val = user_beta
    if beta_val <= 0: calculation_log.append(f"⚠️ Warning: Non-positive Beta ({beta_val:.2f}).")
    components['beta'] = beta_val; components['risk_free_rate'] = user_rf; components['market_risk_premium'] = user_mrp
    cost_of_equity = user_rf + beta_val * user_mrp
    components['cost_of_equity'] = cost_of_equity
    calculation_log.append(f"Inputs -> Rf: {format_percent(user_rf)}, Beta: {beta_val:.2f}, MRP: {format_percent(user_mrp)}")
    calculation_log.append(f"↳ Cost of Equity (Ke) [CAPM]: {format_percent(cost_of_equity)}")

    cost_of_debt_pre_tax = np.nan; tax_rate = np.nan; after_tax_cost_of_debt = np.nan
    total_debt = 0.0; debt_calculation_method = "N/A"; tax_rate_source = "N/A"

    if not balance_sheet_df.empty and not financials_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]; latest_fin = financials_df.iloc[:, 0]
        debt_keys = ['Total Debt', 'Long Term Debt', 'Long Term Debt And Capital Lease Obligation', 'Total Debt Non Current']
        long_term_debt = safe_get(latest_bs, debt_keys , 'Balance Sheet', default=0)
        current_debt_keys = ['Current Debt', 'Short Term Debt', 'Current Debt And Capital Lease Obligation', 'Short Long Term Debt', 'Total Debt Current']
        short_term_debt = safe_get(latest_bs, current_debt_keys , 'Balance Sheet', default=0)
        total_debt = long_term_debt + short_term_debt
        components['debt_value_used'] = total_debt
        interest_expense = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], 'Income Statement', default=0))

        implied_rate_calculated = False
        if total_debt > 1000 and interest_expense > 0:
             implied_rate = interest_expense / total_debt; implied_rate_calculated = True
             if MIN_COST_OF_DEBT <= implied_rate <= MAX_COST_OF_DEBT:
                  cost_of_debt_pre_tax = implied_rate; debt_calculation_method = "Implied";
                  calculation_log.append(f"Interest Exp: {format_currency(interest_expense, symbol=currency_sym, is_indian=is_indian)}, Debt: {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
                  calculation_log.append(f"↳ Implied Cost of Debt (Kd pre-tax): {format_percent(cost_of_debt_pre_tax)}")
             else: calculation_log.append(f"⚠️ Implied Kd ({format_percent(implied_rate)}) out of bounds. Falling back.")
        elif total_debt > 1000: calculation_log.append(f"ℹ️ Found Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}) but no Interest Exp. Falling back.")

        if not debt_calculation_method.startswith("Implied"):
             if total_debt > 1000:
                 debt_calculation_method = "Default (Rf+Spread)"
                 if user_credit_spread is None:
                     spread_prompt = f"Implied Kd failed/NA. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
                     user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
                     user_credit_spread = user_credit_spread_pct / 100.0
                     input_sources_dict['Credit Spread'] = 'User Input (Default Kd)'
                 else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
                 cost_of_debt_pre_tax = user_rf + user_credit_spread
                 calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(cost_of_debt_pre_tax)}")
                 calculation_log.append(f"   (Spread Source: {input_sources_dict['Credit Spread']})")
             else:
                 debt_calculation_method = "Zero Debt"; cost_of_debt_pre_tax = 0.0; input_sources_dict['Credit Spread'] = 'N/A (Zero Debt)';
                 calculation_log.append(f"ℹ️ Negligible Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}). Kd = 0%.")

        income_tax_expense = safe_get(latest_fin, ['Tax Provision', 'Income Tax Expense'], 'Income Statement', default=0)
        pretax_income = safe_get(latest_fin, ['Pretax Income', 'EBT', 'Earnings Before Tax'], 'Income Statement', default=0)
        if pretax_income <= 0 and income_tax_expense > 0:
             ebit_val = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=0); interest_exp_for_ebt = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], default=0))
             pretax_income_from_ebit = ebit_val - interest_exp_for_ebt
             if pretax_income_from_ebit > 0: pretax_income = pretax_income_from_ebit
        if pretax_income > 0 and income_tax_expense >= 0:
            calculated_tax_rate = income_tax_expense / pretax_income; tax_rate = min(max(calculated_tax_rate, MIN_TAX_RATE), MAX_TAX_RATE); tax_rate_source = "Derived & Bounded"
            if abs(tax_rate - calculated_tax_rate) > 0.001: tax_rate_source = f"Derived ({format_percent(calculated_tax_rate)}), Bounded"
            calculation_log.append(f"Tax Exp: {format_currency(income_tax_expense, symbol=currency_sym, is_indian=is_indian)}, Pretax Inc: {format_currency(pretax_income, symbol=currency_sym, is_indian=is_indian)}")
            calculation_log.append(f"↳ Effective Tax Rate: {format_percent(tax_rate)} ({tax_rate_source})")
        else: tax_rate = DEFAULT_TAX_RATE; tax_rate_source = "Default"; calculation_log.append(f"⚠️ Cannot derive tax rate. Using default: {format_percent(tax_rate)}")
    else: # Financial statements empty
         calculation_log.append("⚠️ Financial statements empty. Using defaults."); debt_calculation_method = "Default (No Fin.)"; tax_rate_source = "Default (No Fin.)"
         if user_credit_spread is None:
              spread_prompt = f"No financials. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
              user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
              user_credit_spread = user_credit_spread_pct / 100.0; input_sources_dict['Credit Spread'] = 'User Input (No Fin.)'
         else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
         cost_of_debt_pre_tax = user_rf + user_credit_spread; tax_rate = DEFAULT_TAX_RATE; total_debt = 0; components['debt_value_used'] = 0
         calculation_log.append(f"Using Default Kd: {format_percent(cost_of_debt_pre_tax)} (Rf + Spread: {input_sources_dict['Credit Spread']})")
         calculation_log.append(f"Using Default Tax Rate: {format_percent(tax_rate)}")

    components['cost_of_debt'] = cost_of_debt_pre_tax if np.isfinite(cost_of_debt_pre_tax) else 0.0
    components['debt_calc_method'] = debt_calculation_method
    components['tax_rate'] = tax_rate if np.isfinite(tax_rate) else DEFAULT_TAX_RATE
    components['tax_rate_source'] = tax_rate_source
    if np.isfinite(cost_of_debt_pre_tax) and np.isfinite(tax_rate): after_tax_cost_of_debt = cost_of_debt_pre_tax * (1 - tax_rate)
    else: after_tax_cost_of_debt = components['cost_of_debt'] * (1 - DEFAULT_TAX_RATE) # Use default if calculation failed
    components['after_tax_cost_of_debt'] = after_tax_cost_of_debt
    if components['cost_of_debt'] > 0 : calculation_log.append(f"↳ Cost of Debt (After-tax): {format_percent(after_tax_cost_of_debt)}")

    market_cap_val = safe_get(stock_info_dict, ['marketCap'], 'Info', default=np.nan); equity_source = "Market Cap" if np.isfinite(market_cap_val) and market_cap_val > 0 else "N/A"
    equity_value_for_wacc = np.nan
    if np.isfinite(market_cap_val) and market_cap_val > 0: equity_value_for_wacc = market_cap_val
    else:
       calculation_log.append("⚠️ Market Cap unavailable/zero."); equity_source = "Book Equity Fallback"
       if not balance_sheet_df.empty:
           equity_keys = ['Total Stockholder Equity', 'Stockholders Equity', 'Total Equity Gross Minority Interest', 'Common Stock Equity']
           book_equity_val = safe_get(balance_sheet_df.iloc[:, 0], equity_keys, 'Balance Sheet', default=np.nan)
           if np.isfinite(book_equity_val) and book_equity_val > 1.0:
              equity_value_for_wacc = book_equity_val; calculation_log.append(f"Using Book Equity ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}) as proxy.") # Pass flag
           else: equity_source = "Placeholder (Book Invalid)"
       else: equity_source = "Placeholder (No BS)"
       if not np.isfinite(equity_value_for_wacc): equity_value_for_wacc = 1.0; calculation_log.append("⚠️ Market Cap & Book Equity unavailable. Using placeholder Equity Value (1.0). Weights inaccurate.")
    components['equity_value_used'] = equity_value_for_wacc; components['equity_source'] = equity_source

    equity_weight = np.nan; debt_weight = np.nan; weights_source = "N/A"; total_capital = equity_value_for_wacc + total_debt
    if np.isfinite(total_capital) and total_capital > 1.0:
        equity_weight = equity_value_for_wacc / total_capital; debt_weight = total_debt / total_capital; weights_source = "Calculated"
        total_weight = equity_weight + debt_weight # Normalize if calculation slightly off
        if abs(total_weight - 1.0) > 1e-6: equity_weight /= total_weight; debt_weight /= total_weight
    else: calculation_log.append(f"❌ ERROR: Total Capital ({format_currency(total_capital, symbol=currency_sym, is_indian=is_indian)}) invalid. Using default weights (70/30)."); equity_weight = 0.7; debt_weight = 0.3; weights_source = "Default" # Pass flag
    components['equity_weight'] = equity_weight; components['debt_weight'] = debt_weight; components['weights_source'] = weights_source
    calculation_log.append(f"Equity Value Used ({components['equity_source']}): {format_currency(components['equity_value_used'], symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"Debt Value Used ({components['debt_calc_method']}): {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"↳ Weights E/V: {format_percent(equity_weight)}, D/V: {format_percent(debt_weight)} (Source: {weights_source})")

    wacc = np.nan
    if all(np.isfinite(v) for v in [equity_weight, cost_of_equity, debt_weight, after_tax_cost_of_debt]):
        wacc = (equity_weight * cost_of_equity) + (debt_weight * after_tax_cost_of_debt); original_wacc = wacc; wacc = min(max(wacc, MIN_WACC), MAX_WACC)
        bound_note = ""
        if abs(wacc - original_wacc) > 0.0001 : bound_note = f"(Bounded from {format_percent(original_wacc)})"; calculation_log.append(f"⚠️ WACC ({format_percent(original_wacc)}) bounded.")
        calculation_log.append(f"▶️ Calculated WACC: {format_percent(wacc)} {bound_note}")
    else: calculation_log.append(f"❌ ERROR: Cannot calculate final WACC due to invalid components."); wacc = np.nan
    components['final_wacc'] = wacc

    display(Markdown("```\n" + "\n".join(calculation_log) + "\n```"))

    if np.isfinite(wacc) and wacc > 0 and np.isfinite(equity_weight) and np.isfinite(debt_weight):
        equity_contrib = equity_weight * cost_of_equity; debt_contrib = debt_weight * after_tax_cost_of_debt
        if np.isfinite(equity_contrib) and np.isfinite(debt_contrib) and abs(equity_contrib + debt_contrib - wacc) < 0.01:
           labels = ['Cost of Equity Contribution', 'Cost of Debt Contribution']; values = [equity_contrib, debt_contrib]
           fig_wacc_donut = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4, marker_colors=px.colors.qualitative.Pastel, hovertemplate='<b>%{label}</b><br>Contribution: %{value:.2%}<extra></extra>', textinfo='percent', textfont_size=12, sort=False)])
           fig_wacc_donut.update_layout(title_text=f'WACC Composition (Total: {format_percent(wacc)})', title_x=0.5, showlegend=True, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), margin=dict(l=20, r=20, t=60, b=50), height=350)
           fig_wacc_donut.show()
        else: display(Markdown(f"*(WACC component contributions invalid/don't sum. Cannot generate chart.)*"))
    elif not np.isfinite(wacc): display(Markdown(f"*(WACC calculation failed. Cannot generate chart)*"))
    else: display(Markdown(f"*(WACC zero/negative or weights invalid. Cannot generate chart)*"))

    return wacc, components

# Execute WACC Calculation
discount_rate, wacc_components = calculate_wacc(
    stock_data['info'], stock_data['financials'], stock_data['balance_sheet'],
    risk_free_rate_input, market_risk_premium_input, beta_input, input_sources,
    currency_symbol, is_indian_stock
)

# %% [markdown]
# ## 6. FCFF Calculation & Projection (Initial Growth: Resilient GM of valid 3yr CAGRs)

# %%
# FCFF Function Definition (NEW initial growth logic - V14 - Resilient Projection)
def calculate_and_project_fcff(financials_df, balance_sheet_df, cashflow_df, history_long_df, # History needed for stock CAGR
                               wacc_comps, disc_rate, forecast_yrs,
                               final_user_rf, default_init_growth, # This is 6.5% default
                               currency_sym, is_indian): # Added currency params
    display(Markdown("### Calculating & Projecting FCFF..."))
    projection_log = []
    projection_results = {
        'historical_fcffs': {}, # Store recent historical FCFFs
        'historical_revenues': {}, # Store recent historical revenues
        'current_fcff': np.nan, 'projected_fcffs': [], 'growth_rates': [],
        'pv_fcffs': [], 'cumulative_pv_fcff': np.nan, 'terminal_growth_rate': np.nan,
        'terminal_value': np.nan, 'pv_terminal_value': np.nan, 'error': None,
        'initial_growth_rate': np.nan, 'growth_source': 'N/A'
    }

    if financials_df.empty or balance_sheet_df.empty or cashflow_df.empty:
         projection_results['error'] = "Missing Financials/Balance Sheet/Cash Flow."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results
    if not np.isfinite(disc_rate):
         projection_results['error'] = "Invalid WACC provided."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results

    # --- Define Helper to calculate FCFF for a specific year column ---
    def get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, tax_rt):
        if not isinstance(fin_col, pd.Series) or not isinstance(bs_col, pd.Series) or not isinstance(cf_col, pd.Series): return np.nan, "Missing Statement Column"
        ebit = safe_get(fin_col, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
        depreciation_amortization = safe_get(cf_col, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
        capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
        capital_expenditures = abs(safe_get(cf_col, capex_keys, default=0))
        wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]
        change_in_wc = safe_get(cf_col, wc_keys, default=None) ; wc_src = "CF"
        if change_in_wc is None or not np.isfinite(change_in_wc):
            wc_src = "BS Calc"
            if isinstance(bs_prev_col, pd.Series):
                 curr_assets = safe_get(bs_col, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(bs_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                 prev_assets = safe_get(bs_prev_col, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(bs_prev_col, ['Current Liabilities', 'Total Current Liabilities'], default=0)
                 prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                 if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val): change_in_wc = curr_wc_val - prev_wc_val
                 else: change_in_wc = 0; wc_src = "BS Fail"
            else: change_in_wc = 0; wc_src = "No Prev BS"
        if not np.isfinite(change_in_wc): change_in_wc = 0
        fcff_val = np.nan; calc_status = "OK"
        if np.isfinite(ebit) and np.isfinite(tax_rt):
            nopat_val = ebit * (1 - tax_rt)
            if np.isfinite(depreciation_amortization) and np.isfinite(capital_expenditures) and np.isfinite(change_in_wc): fcff_val = nopat_val + depreciation_amortization - capital_expenditures - change_in_wc
            else: calc_status = "Invalid Depr/CapEx/WC"
        else: calc_status = "Invalid EBIT/Tax"
        return fcff_val, calc_status
    # --- End FCFF Helper ---

    # --- Calculate Historical FCFF & Revenue (need T-0 and T-3 for CAGR) ---
    num_years_cagr = 3
    max_hist_years = num_years_cagr + 1 # Need 4 data points for 3yr CAGR
    num_statement_years = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    historical_fcffs_list = [] # Newest first (index 0 = T-0)
    historical_revenues_list = [] # Newest first
    # Use WACC tax rate for consistency, or default if WACC failed
    effective_tax_rate = DEFAULT_TAX_RATE
    if isinstance(wacc_comps, dict) and 'tax_rate' in wacc_comps and np.isfinite(wacc_comps['tax_rate']):
        effective_tax_rate = wacc_comps['tax_rate']

    projection_log.append(f"Calculating historical FCFF & Revenue (up to {max_hist_years} points)...")
    years_to_calc = min(num_statement_years, max_hist_years)
    for i in range(years_to_calc):
        fin_col = financials_df.iloc[:, i]; bs_col = balance_sheet_df.iloc[:, i]; cf_col = cashflow_df.iloc[:, i]
        bs_prev_col = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None
        fcff_val, status = get_fcff_for_year(fin_col, bs_col, cf_col, bs_prev_col, effective_tax_rate)
        historical_fcffs_list.append(fcff_val if status == "OK" else np.nan)
        rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
        revenue_val = safe_get(fin_col, rev_keys, default=np.nan)
        historical_revenues_list.append(revenue_val)

    projection_results['historical_fcffs'] = {f"T-{i}": v for i, v in enumerate(historical_fcffs_list)}
    projection_results['historical_revenues'] = {f"T-{i}": v for i, v in enumerate(historical_revenues_list)}
    fcff_current = historical_fcffs_list[0] if historical_fcffs_list else np.nan # Get T-0 FCFF (might be NaN)
    projection_results['current_fcff'] = fcff_current

    # --- Display Calculation Log for Current FCFF ---
    # Log the result even if it's invalid, but don't make it sound like a success
    tax_rate_source_disp = wacc_comps.get('tax_rate_source','?') if isinstance(wacc_comps, dict) else '?'
    if historical_fcffs_list: # Check if list is not empty
        if np.isfinite(fcff_current):
            # Code to get components for logging (same as before)
            latest_fin = financials_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]
            ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
            nopat = ebit * (1-effective_tax_rate) if np.isfinite(ebit) else np.nan
            depreciation_amortization = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0)
            capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]; capital_expenditures = abs(safe_get(latest_cf, capex_keys, default=0))
            wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]; change_in_wc = safe_get(latest_cf, wc_keys, default=None); wc_source = "Direct from Cash Flow"
            if change_in_wc is None or not np.isfinite(change_in_wc):
                if balance_sheet_df.shape[1] >= 2:
                    prev_bs = balance_sheet_df.iloc[:, 1]
                    curr_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=0); curr_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0); curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
                    prev_assets = safe_get(prev_bs, ['Current Assets', 'Total Current Assets'], default=0); prev_liab = safe_get(prev_bs, ['Current Liabilities', 'Total Current Liabilities'], default=0); prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
                    if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val): change_in_wc = curr_wc_val - prev_wc_val; wc_source = "Calculated from BS"
                    else: change_in_wc = 0; wc_source = "Calc Failed (Invalid BS)"
                else: change_in_wc = 0; wc_source = "Default (No Prev BS)"
            if not np.isfinite(change_in_wc): change_in_wc = 0
            projection_log.append(f"EBIT (T-0): {format_currency(ebit, currency_sym, is_indian=is_indian)}, Tax Rate ({tax_rate_source_disp}) {format_percent(effective_tax_rate)} -> NOPAT: {format_currency(nopat, currency_sym, is_indian=is_indian)}")
            projection_log.append(f"  + Dep&Amort: {format_currency(depreciation_amortization, currency_sym, is_indian=is_indian)}")
            projection_log.append(f"  - CapEx: {format_currency(capital_expenditures, currency_sym, is_indian=is_indian)}")
            projection_log.append(f"  - Change WC ({wc_source}): {format_currency(change_in_wc, currency_sym, is_indian=is_indian)}")
            projection_log.append(f"▶️ Calculated Historical FCFF (T-0): {format_currency(fcff_current, currency_sym, is_indian=is_indian)}")
        else:
             projection_log.append(f"⚠️ Could not calculate valid historical FCFF (T-0). Value: {fcff_current}")
    else:
        projection_log.append(f"⚠️ No historical FCFF data available.")


    # --- V14: REMOVED early exit based on fcff_current validity ---
    # The growth rate calculation will proceed using available data.

    # --- Initial Growth Rate Calculation (GM of valid 3yr CAGRs - V14) ---
    projection_log.append("--- Determining Initial Growth Rate (Geometric Mean of valid 3yr CAGRs) ---")
    initial_growth_rate = np.nan # Start with NaN
    growth_source = "N/A"
    calculated_cagrs = {} # Store individual CAGRs {Name: value}
    valid_cagr_list = [] # Store only the valid CAGRs for GM calculation

    # 1. Calculate 3-Year Revenue CAGR (Conditional)
    rev_cagr = np.nan
    projection_log.append(f"Calculating Revenue {num_years_cagr}yr CAGR...")
    if len(historical_revenues_list) >= max_hist_years:
        rev_end = historical_revenues_list[0]           # T-0
        rev_start = historical_revenues_list[num_years_cagr] # T-3
        if pd.notna(rev_end) and math.isfinite(rev_end) and pd.notna(rev_start) and math.isfinite(rev_start) and rev_start > 1e-9:
            rev_cagr = calculate_cagr_modified(rev_start, rev_end, num_years_cagr)
            if pd.notna(rev_cagr) and math.isfinite(rev_cagr):
                 calculated_cagrs['Revenue'] = rev_cagr; valid_cagr_list.append(rev_cagr)
                 projection_log.append(f"  ✅ Revenue CAGR: {format_percent(rev_cagr)} (Start: {format_currency(rev_start, '', 0, is_indian)}, End: {format_currency(rev_end, '', 0, is_indian)})")
            else: projection_log.append(f"  ❌ Revenue CAGR Calculation Failed (Result Invalid). (S={rev_start:.0f}, E={rev_end:.0f})")
        else: projection_log.append(f"  ❌ Revenue CAGR Skipped (Invalid/Non-positive Start Data). (S={rev_start}, E={rev_end})")
    else: projection_log.append(f"  ❌ Revenue CAGR Skipped (Insufficient data: {len(historical_revenues_list)} points)")

    # 2. Calculate 3-Year FCFF CAGR (Conditional)
    fcff_cagr = np.nan
    projection_log.append(f"Calculating FCFF {num_years_cagr}yr CAGR...")
    if len(historical_fcffs_list) >= max_hist_years:
        fcff_end = historical_fcffs_list[0]           # T-0
        fcff_start = historical_fcffs_list[num_years_cagr] # T-3
        if pd.notna(fcff_end) and math.isfinite(fcff_end) and pd.notna(fcff_start) and math.isfinite(fcff_start) and fcff_start > 1e-9:
            fcff_cagr = calculate_cagr_modified(fcff_start, fcff_end, num_years_cagr)
            if pd.notna(fcff_cagr) and math.isfinite(fcff_cagr):
                 calculated_cagrs['FCFF'] = fcff_cagr; valid_cagr_list.append(fcff_cagr)
                 projection_log.append(f"  ✅ FCFF CAGR:    {format_percent(fcff_cagr)} (Start: {format_currency(fcff_start, '', 0, is_indian)}, End: {format_currency(fcff_end, '', 0, is_indian)})")
            else: projection_log.append(f"  ❌ FCFF CAGR Calculation Failed (Result Invalid). (S={fcff_start:.0f}, E={fcff_end:.0f})")
        else: projection_log.append(f"  ❌ FCFF CAGR Skipped (Invalid/Non-positive Start Data). (S={fcff_start}, E={fcff_end})")
    else: projection_log.append(f"  ❌ FCFF CAGR Skipped (Insufficient data: {len(historical_fcffs_list)} points)")

    # 3. Calculate 3-Year Stock Price CAGR (Conditional - Robust Column V14)
    stock_cagr = np.nan
    price_col = None # Determine which price column to use
    projection_log.append(f"Calculating Stock Price {num_years_cagr}yr CAGR...")
    if history_long_df is not None and not history_long_df.empty and len(history_long_df) >= 2:
         if 'Adj Close' in history_long_df.columns: price_col = 'Adj Close'
         elif 'Close' in history_long_df.columns: price_col = 'Close'; projection_log.append("  ⚠️ Using 'Close' price for Stock CAGR ('Adj Close' not found).")
         else: projection_log.append(f"  ❌ Stock CAGR Skipped (Neither 'Adj Close' nor 'Close' columns found)")

         if price_col and isinstance(history_long_df.index, pd.DatetimeIndex):
             projection_log.append(f"  Stock History Info: Using '{price_col}' column. Found {len(history_long_df)} rows, Date Range: {history_long_df.index.min().date()} to {history_long_df.index.max().date()}")
             try:
                 end_date = history_long_df.index[-1]; end_price = history_long_df[price_col].iloc[-1]
                 start_date_target = end_date - pd.DateOffset(years=num_years_cagr)
                 # Find index >= target date
                 temp_series = history_long_df.loc[start_date_target:]
                 if not temp_series.empty:
                      start_actual_date = temp_series.index[0]
                      start_price = temp_series[price_col].iloc[0]
                      actual_years = (end_date - start_actual_date).days / 365.25

                      projection_log.append(f"  Stock CAGR Data Points: Start Date={start_actual_date.date()}, Start Price={start_price:.2f}, End Date={end_date.date()}, End Price={end_price:.2f}, Actual Years={actual_years:.2f}")

                      is_end_valid = pd.notna(end_price) and math.isfinite(end_price)
                      is_start_valid = pd.notna(start_price) and math.isfinite(start_price) and start_price > 1e-9
                      is_period_valid = actual_years >= (num_years_cagr - 0.5) and actual_years > 0

                      if is_end_valid and is_start_valid and is_period_valid:
                          stock_cagr = calculate_cagr_modified(start_price, end_price, actual_years)
                          if pd.notna(stock_cagr) and math.isfinite(stock_cagr):
                              calculated_cagrs['Stock'] = stock_cagr; valid_cagr_list.append(stock_cagr)
                              projection_log.append(f"  ✅ Stock CAGR:   {format_percent(stock_cagr)} (Using '{price_col}')")
                          else: projection_log.append(f"  ❌ Stock CAGR Calculation Failed (Result Invalid).")
                      else:
                          reason = [];
                          if not is_end_valid: reason.append("Invalid End Price");
                          if not is_start_valid: reason.append("Invalid/Non-positive Start Price")
                          if not is_period_valid: reason.append(f"Period Invalid ({actual_years:.2f} yrs vs target {num_years_cagr})")
                          projection_log.append(f"  ❌ Stock CAGR Skipped ({', '.join(reason)}).")
                 else:
                     projection_log.append(f"  ❌ Stock CAGR Skipped (Could not find data at or after target start date: {start_date_target.date()})")
             except Exception as e:
                  projection_log.append(f"  ❌ Stock CAGR Skipped (Error during lookup/calculation: {e})")
         elif not price_col: pass # Already logged the reason
         else: projection_log.append(f"  ❌ Stock CAGR Skipped (History index is not DatetimeIndex)")
    else: # history_long_df checks failed
         reason = [];
         if history_long_df is None: reason.append("History DF is None")
         elif history_long_df.empty: reason.append("History DF is empty")
         elif len(history_long_df) < 2: reason.append("Less than 2 data points")
         projection_log.append(f"  ❌ Stock CAGR Skipped (Insufficient/invalid history data object: {', '.join(reason)})")


    # 4. Calculate Geometric Mean of the valid CAGRs
    if valid_cagr_list:
        gm_growth = geometric_mean_modified(valid_cagr_list)
        if pd.notna(gm_growth) and math.isfinite(gm_growth):
            initial_growth_rate = gm_growth
            growth_source = f"GM of {len(valid_cagr_list)} valid CAGR(s)"
            projection_log.append(f"  ▶️ Calculated Geometric Mean = {format_percent(initial_growth_rate)}")
            included_cagrs = [f"{k}:{format_percent(v)}" for k, v in calculated_cagrs.items()] # Show all attempted/successful
            projection_log.append(f"     Based on: [{', '.join(included_cagrs)}]")
        else:
            projection_log.append(f"  ⚠️ Geometric Mean calculation failed or result invalid. Falling back to default.")
            initial_growth_rate = default_init_growth
            growth_source = f"Default ({format_percent(default_init_growth)}) [GM Fail]"
            projection_log.append(f"     CAGRs calculated but GM failed: {valid_cagr_list}") # Log the rates that caused GM failure
    else:
        # If NO valid CAGRs were calculated
        projection_log.append(f"  ⚠️ No valid 3yr CAGRs could be calculated. Falling back to default.")
        initial_growth_rate = default_init_growth
        growth_source = f"Default ({format_percent(default_init_growth)}) [No Valid CAGRs]"

    # 5. Apply Floor (1.1%) and Ceiling (15%)
    unbounded_rate = initial_growth_rate # Store original for comparison
    initial_growth_rate = max(initial_growth_rate, MIN_INITIAL_GROWTH_FLOOR) # Apply 1.1% floor
    initial_growth_rate = min(initial_growth_rate, MAX_INITIAL_GROWTH_CEILING) # Apply 15% ceiling

    if abs(initial_growth_rate - unbounded_rate) > 1e-6: # Check if bounding changed the value
        growth_source += " (Bounded)"
        projection_log.append(f"  Applied Bounds: Final Initial Growth Rate = {format_percent(initial_growth_rate)} (Floor: {format_percent(MIN_INITIAL_GROWTH_FLOOR)}, Ceiling: {format_percent(MAX_INITIAL_GROWTH_CEILING)})")
    else:
         projection_log.append(f"  Final Initial Growth Rate = {format_percent(initial_growth_rate)}")

    projection_results['initial_growth_rate'] = initial_growth_rate
    projection_results['growth_source'] = growth_source

    # --- V14: Check FCFF T-0 validity BEFORE projecting ---
    if not np.isfinite(fcff_current):
        error_msg = "Invalid historical FCFF (T-0). Cannot proceed with FCFF projection."
        projection_results['error'] = error_msg
        projection_log.append(f"❌ ERROR: {error_msg}")
        display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
        return projection_results # Stop before projection


    # --- Terminal Growth Rate Calculation (Same conservative approach) ---
    g_cap_initial_valid = initial_growth_rate if pd.notna(initial_growth_rate) and math.isfinite(initial_growth_rate) else TERMINAL_GROWTH_RATE_CAP # Fallback if initial is invalid
    g_cap_config = TERMINAL_GROWTH_RATE_CAP; g_cap_wacc = disc_rate * 0.67; g_cap_rf = final_user_rf * 0.95 # Use final RF
    projection_log.append(f"--- Determining Terminal Growth Rate ---")
    projection_log.append(f"Terminal Growth Caps Considered: Config={format_percent(g_cap_config)}, WACC*0.67={format_percent(g_cap_wacc)}, Initial(Final)={format_percent(g_cap_initial_valid)}, Rf*0.95={format_percent(g_cap_rf)}")
    terminal_growth_rate = min(g_cap_config, g_cap_wacc, g_cap_initial_valid, g_cap_rf); projection_log.append(f"↳ Minimum of Caps: {format_percent(terminal_growth_rate)}")
    original_g_before_floor = terminal_growth_rate
    terminal_growth_rate = max(MIN_GROWTH, terminal_growth_rate) # Apply floor (MIN_GROWTH=0.00 usually)
    if abs(terminal_growth_rate - original_g_before_floor) > 1e-6 : projection_log.append(f"↳ Adjusted by Floor ({format_percent(MIN_GROWTH)}): {format_percent(terminal_growth_rate)}")
    if terminal_growth_rate >= disc_rate: original_g = terminal_growth_rate; terminal_growth_rate = disc_rate - 0.005; projection_log.append(f"⚠️ Terminal Growth ({format_percent(original_g)}) >= WACC ({format_percent(disc_rate)}). Reduced to {format_percent(terminal_growth_rate)}.")
    projection_log.append(f"▶️ Using Final Terminal Growth Rate (g): {format_percent(terminal_growth_rate)}"); projection_results['terminal_growth_rate'] = terminal_growth_rate

    # --- Project FCFFs using Linear Decay from initial_growth_rate -> terminal_growth_rate ---
    projected_fcffs = []; growth_rates_applied = []; pv_fcffs = []; cumulative_pv_fcff = 0.0
    current_fcff_for_proj = fcff_current # Start projection base from T-0 FCFF (Now guaranteed to be finite)

    # Ensure effective initial growth is valid (should be after default/bounding)
    if pd.isna(initial_growth_rate) or not math.isfinite(initial_growth_rate):
        projection_log.append(f"⚠️ FATAL: Invalid final initial growth rate ({initial_growth_rate}) despite checks. Using 0% for projection start.")
        effective_initial_growth = 0.0
    else:
        effective_initial_growth = initial_growth_rate # Use the final bounded rate

    effective_terminal_growth = terminal_growth_rate

    # Handle negative starting FCFF carefully (although fcff_current is now finite, it could still be negative)
    if current_fcff_for_proj <= 0:
        projection_log.append("⚠️ Starting FCFF (T-0) <= 0. Applying minimal growth.")
        effective_initial_growth = min(effective_initial_growth, 0.01);
        effective_terminal_growth = min(terminal_growth_rate, 0.00)

    projection_log.append("--- FCFF Projections ---"); projection_log.append(f"{'Year':<5} {'Growth':<8} {'FCFF':<18} {'PV Factor':<10} {'PV FCFF':<18}"); projection_log.append("-" * 60)
    last_valid_fcff = current_fcff_for_proj

    for year in range(1, forecast_yrs + 1):
        if forecast_yrs > 0: decay_factor = max(0, (forecast_yrs - year)) / forecast_yrs
        else: decay_factor = 0
        growth_rate = effective_terminal_growth + (effective_initial_growth - effective_terminal_growth) * decay_factor
        growth_rate = max(MIN_GROWTH, growth_rate) # Apply general floor (0%)
        if current_fcff_for_proj <= 0: growth_rate = max(-0.10, min(growth_rate, 0.05)) # Moderate if start neg

        growth_rates_applied.append(growth_rate)
        fcff_forecast = last_valid_fcff * (1 + growth_rate)
        if not np.isfinite(fcff_forecast): fcff_forecast = 0
        projected_fcffs.append(fcff_forecast)
        last_valid_fcff = fcff_forecast # Update last_valid_fcff

        pv_factor = 1 / ((1 + disc_rate) ** year); pv_fcff = fcff_forecast * pv_factor
        if not np.isfinite(pv_fcff): pv_fcff = 0
        pv_fcffs.append(pv_fcff); cumulative_pv_fcff += pv_fcff
        projection_log.append(f"{year:<5} {format_percent(growth_rate):<8} {format_currency(fcff_forecast, '', 0, is_indian=is_indian):<18} {pv_factor:<10.4f} {format_currency(pv_fcff, '', 0, is_indian=is_indian):<18}")

    projection_results['projected_fcffs'] = projected_fcffs; projection_results['growth_rates'] = growth_rates_applied; projection_results['pv_fcffs'] = pv_fcffs
    projection_results['cumulative_pv_fcff'] = cumulative_pv_fcff if np.isfinite(cumulative_pv_fcff) else 0.0
    projection_log.append("-" * 60); projection_log.append(f"Sum of PV(Forecasted FCFFs): {format_currency(cumulative_pv_fcff, symbol=currency_sym, is_indian=is_indian)}")

    # Calculate Terminal Value (Remains the same)
    terminal_value = np.nan; pv_terminal_value = np.nan
    final_forecast_fcff = projected_fcffs[-1] if projected_fcffs and np.isfinite(projected_fcffs[-1]) else 0

    if (disc_rate - terminal_growth_rate) <= 1e-6 or not np.isfinite(terminal_growth_rate):
         projection_log.append(f"❌ ERROR: WACC ({format_percent(disc_rate)}) <= Terminal Growth ({format_percent(terminal_growth_rate)}) or g invalid. TV calc failed."); projection_results['error'] = "WACC <= g or g invalid. TV Calc failed."
    elif not np.isfinite(final_forecast_fcff):
         # This case should be less likely now as fcff_current is checked earlier
         projection_log.append(f"❌ ERROR: Final projected FCFF (Year {forecast_yrs}) is invalid. TV calc failed."); projection_results['error'] = "Final FCFF invalid. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    else:
         if final_forecast_fcff <= 0:
             projection_log.append(f"ℹ️ FCFF Year {forecast_yrs} ({format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)}) <= 0. Setting TV=0.");
             terminal_value = 0.0; pv_terminal_value = 0.0
         else:
              fcff_n_plus_1 = final_forecast_fcff * (1 + terminal_growth_rate)
              terminal_value = fcff_n_plus_1 / (disc_rate - terminal_growth_rate)
              if not np.isfinite(terminal_value) or terminal_value < 0:
                  projection_log.append(f"⚠️ Calculated TV invalid/negative ({format_currency(terminal_value, currency_sym, is_indian=is_indian)}). Setting TV=0.")
                  terminal_value = 0.0; pv_terminal_value = 0.0
              else:
                   pv_factor_tv = 1 / ((1 + disc_rate) ** forecast_yrs); pv_terminal_value = terminal_value * pv_factor_tv
                   projection_log.append(f"FCFF Yr {forecast_yrs}: {format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)} -> FCFF Yr {forecast_yrs+1}: {format_currency(fcff_n_plus_1, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"Terminal Value (Yr {forecast_yrs}): {format_currency(terminal_value, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"PV Factor TV: {pv_factor_tv:.4f} -> PV(TV): {format_currency(pv_terminal_value, currency_sym, is_indian=is_indian)}")

    projection_results['terminal_value'] = terminal_value if np.isfinite(terminal_value) else 0.0
    projection_results['pv_terminal_value'] = pv_terminal_value if np.isfinite(pv_terminal_value) else 0.0

    display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
    return projection_results

# --- Execute FCFF calculation and projection ---
# Initialize fcff_results dictionary to ensure it's defined before the try block
fcff_results = {'error': 'Calculation not run yet.', 'initial_growth_rate': np.nan, 'growth_source':'N/A', 'terminal_growth_rate': np.nan}

if np.isfinite(discount_rate):
    try:
        # Pass currency symbol, flag, default growth rate, AND long history
        fcff_results = calculate_and_project_fcff(
            stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'], stock_data['history_long'], # Pass history_long
            wacc_components, discount_rate, FORECAST_YEARS, risk_free_rate_input, DEFAULT_GROWTH_RATE,
            currency_symbol, is_indian_stock)
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during FCFF Projection: {e}**"))
         # Ensure fcff_results is still a dict with error info
         fcff_results = {'error': f'Runtime error in FCFF calc: {e}', 'initial_growth_rate': np.nan, 'growth_source':'Error', 'terminal_growth_rate': np.nan}

else:
    display(Markdown("❌ **Skipping FCFF Projection due to invalid WACC.**"));
    # Ensure fcff_results is a dict even if skipped
    fcff_results = {'error': 'Invalid WACC.', 'initial_growth_rate': np.nan, 'growth_source':'N/A (WACC Invalid)', 'terminal_growth_rate': np.nan}


# %% [markdown]
# ## 7. Enterprise Value and Equity Value Calculation

# %%
# Valuation Function Definition (No changes needed here, already robust)
def calculate_final_valuation(fcff_data, balance_sheet_df, shares_outstanding_val, currency_sym, wacc_data, is_indian): # Added flag
    display(Markdown("### Final Valuation Calculation..."))
    valuation = {'enterprise_value': np.nan, 'equity_value': np.nan, 'intrinsic_value_per_share': np.nan, 'error': None}
    valuation_log = []

    # Ensure fcff_data is a dictionary before accessing keys
    if not isinstance(fcff_data, dict):
        error_msg = f"FCFF Data is not a dictionary (type: {type(fcff_data)}). Cannot proceed."
        valuation['error'] = error_msg
        valuation_log.append(f"❌ ERROR: {error_msg}")
        display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
        return valuation

    if fcff_data.get('error'):
        valuation['error'] = f"FCFF Error: {fcff_data.get('error', 'Unknown')}"; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation

    pv_forecast_sum = fcff_data.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_data.get('pv_terminal_value', np.nan)
    if not np.isfinite(pv_forecast_sum) or not np.isfinite(pv_tv):
         valuation['error'] = f"Invalid PV components (PV FCFF: {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}, PV TV: {format_currency(pv_tv, currency_sym, is_indian=is_indian)})."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation # Pass flag

    enterprise_value = pv_forecast_sum + pv_tv; valuation['enterprise_value'] = enterprise_value
    valuation_log.append(f"PV(Forecast FCFFs): {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"PV(Terminal Value):   {format_currency(pv_tv, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"↳ Enterprise Value (EV): {format_currency(enterprise_value, currency_sym, is_indian=is_indian)}")

    # Get Debt & Cash from BS (use Debt from WACC for consistency)
    total_debt_latest = 0.0; cash_and_equivalents = 0.0; minority_interest = 0.0
    debt_source_info = '?'
    if isinstance(wacc_data, dict):
        debt_source_info = wacc_data.get('debt_calc_method', '?')
        total_debt_latest = wacc_data.get('debt_value_used', 0.0) # Get debt used in WACC
    else:
        valuation_log.append(f"⚠️ WACC data invalid, cannot get debt used in WACC. Assuming 0.")

    if not balance_sheet_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        cash_keys = ["Cash And Cash Equivalents", "Cash Cash Equivalents And Short Term Investments", "Cash Financial", "Cash", "Cash And Short Term Investments"]
        cash_and_equivalents = safe_get(latest_bs, cash_keys, "BS", default=0)
        valuation_log.append(f"  + Cash & Equiv (BS): {format_currency(cash_and_equivalents, currency_sym, is_indian=is_indian)}")
        minority_interest = safe_get(latest_bs, ["Minority Interest", "Noncontrolling Interest"], "BS", default=0)
        if minority_interest != 0.0:
            valuation_log.append(f"  - Minority Interest: {format_currency(minority_interest, currency_sym, is_indian=is_indian)}")
    else:
        valuation_log.append("⚠️ Balance Sheet empty.")
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        valuation_log.append(f"  + Cash & Equiv: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")

    if not all(np.isfinite(v) for v in [enterprise_value, total_debt_latest, cash_and_equivalents, minority_interest]):
         valuation['error'] = "Invalid component for EV->Equity bridge."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    equity_value = enterprise_value - total_debt_latest + cash_and_equivalents - minority_interest; valuation['equity_value'] = equity_value
    valuation_log.append(f"↳ Equity Value: {format_currency(equity_value, currency_sym, is_indian=is_indian)}")

    # Calculate Intrinsic Value Per Share
    intrinsic_value_per_share = np.nan
    if not np.isfinite(equity_value): valuation['error'] = f"Equity value invalid ({equity_value})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    elif shares_outstanding_val <= 0: valuation['error'] = f"Shares outstanding invalid ({shares_outstanding_val})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    else:
        intrinsic_value_per_share = equity_value / shares_outstanding_val;
        valuation_log.append(f"  ÷ Shares Out: {shares_outstanding_val:,.0f}")
        valuation_log.append(f"▶️ Intrinsic Value/Share: {format_currency(intrinsic_value_per_share, currency_sym, is_indian=is_indian)}")
    valuation['intrinsic_value_per_share'] = intrinsic_value_per_share if np.isfinite(intrinsic_value_per_share) else np.nan

    display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
    return valuation

# --- Execute Final Valuation ---
# Initialize final_valuation to ensure it's defined before the try block
final_valuation = {'error': 'Calculation not run yet.', 'intrinsic_value_per_share': np.nan}

# Check if fcff_results and wacc_components are dictionaries and fcff_results has no error
if isinstance(fcff_results, dict) and isinstance(wacc_components, dict) and not fcff_results.get('error'):
    try:
        final_valuation = calculate_final_valuation(
            fcff_results, stock_data['balance_sheet'], shares_outstanding,
            currency_symbol, wacc_components, is_indian_stock
        )
    except Exception as e:
        display(Markdown(f"❌ **CRITICAL ERROR during Final Valuation: {e}**"))
        final_valuation = {'error': f'Runtime error in final valuation: {e}', 'intrinsic_value_per_share': np.nan}
elif isinstance(fcff_results, dict): # If fcff_results has an error
     display(Markdown("❌ **Skipping Final Valuation due to FCFF errors.**"));
     final_valuation = {'error': fcff_results.get('error', 'Upstream FCFF error.'), 'intrinsic_value_per_share': np.nan}
else: # If fcff_results or wacc_components are not dictionaries
     display(Markdown("❌ **Skipping Final Valuation due to critical upstream errors (invalid data structures).**"));
     final_valuation = {'error': 'FCFF or WACC results dictionary not created/valid.', 'intrinsic_value_per_share': np.nan}


# %% [markdown]
# ## 8. Financial Ratio Analysis

# %%
# Ratio Function Definition (No changes needed here, already robust)
def analyze_ratios(stock_info_dict, financials_df, balance_sheet_df, cashflow_df,
                   enterprise_val, current_price_val, shares_outstanding_val,
                   currency_sym, user_benchmark_dict, # Only user inputs
                   benchmark_source_dict, # Dict tracking source per ratio
                   wacc_data, is_indian): # Pass WACC data for consistent debt and flag

    """ Calculates ratios, compares to user/fallback benchmarks, uses WACC-consistent debt. """
    display(Markdown("### Financial Ratio Analysis..."))
    ratios = {} # Calculated ratios {RatioName: value}
    ratio_log = []
    ratio_data_for_table = [] # For HTML table display
    ratio_plotly_data = {'Ratio': [], 'Company': [], 'Benchmark': [], 'Benchmark Source': [], 'Status': []} # For Plotly chart
    company_nm = stock_info_dict.get("shortName", ticker_symbol)

    # --- Pre-checks ---
    valid_price = current_price_val is not None and np.isfinite(current_price_val) and current_price_val > 0
    valid_shares = shares_outstanding_val is not None and np.isfinite(shares_outstanding_val) and shares_outstanding_val > 0
    valid_enterprise_val = enterprise_val is not None and np.isfinite(enterprise_val)

    # Check for essential statements
    if financials_df.empty or balance_sheet_df.empty:
         ratio_log.append("❌ ERROR: Income Statement or Balance Sheet empty. Cannot calculate most ratios.")
         display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
         return {}, [] # Return empty results

    latest_fin = financials_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0] if not cashflow_df.empty else pd.Series(dtype=float)

    # --- Calculate Ratio Components (using safe_get with NaN default where appropriate) ---
    net_income = safe_get(latest_fin, ["Net Income Continuous Operations", "Net Income", "Net Income Common Stockholders", "Net Income Applicable To Common Shares"], default=np.nan)
    revenue = safe_get(latest_fin, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
    depreciation = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0) # Default 0 ok here
    ebitda = (ebit + depreciation) if np.isfinite(ebit) and np.isfinite(depreciation) else np.nan

    total_equity = safe_get(latest_bs, ["Total Stockholder Equity", "Stockholders Equity", "Total Equity Gross Minority Interest", "Common Stock Equity"], default=np.nan)
    total_assets = safe_get(latest_bs, ["Total Assets"], default=np.nan)
    current_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=np.nan)
    current_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=np.nan)
    inventory = safe_get(latest_bs, ["Inventory"], default=0) # Default 0 ok here

    # --- Use Debt value consistent with WACC calculation ---
    total_debt_ratio = 0.0; debt_source_info = '?'
    if isinstance(wacc_data, dict):
        total_debt_ratio = wacc_data.get('debt_value_used', 0.0)
        debt_source_info = wacc_data.get('debt_calc_method', '?');
    else:
        ratio_log.append(f"⚠️ WACC data invalid, cannot get debt used in WACC. Assuming 0 for ratio calcs.")
    ratio_log.append(f"Using Debt = {format_currency(total_debt_ratio, currency_sym, is_indian=is_indian)} (Source: WACC - {debt_source_info})")

    # Per Share Metrics
    eps = (net_income / shares_outstanding_val) if valid_shares and np.isfinite(net_income) and net_income != 0 else np.nan
    bvps = (total_equity / shares_outstanding_val) if valid_shares and np.isfinite(total_equity) and total_equity != 0 else np.nan

    # Capital Employed (ROCE denominator)
    capital_employed = np.nan; ce_source = "N/A"
    if np.isfinite(total_assets) and np.isfinite(current_liab):
        capital_employed_def1 = total_assets - current_liab
        if np.isfinite(capital_employed_def1) and capital_employed_def1 > 1.0:
            capital_employed = capital_employed_def1; ce_source = "A-CL"
    if not np.isfinite(capital_employed):
        if np.isfinite(total_equity) and np.isfinite(total_debt_ratio):
             capital_employed_def2 = total_equity + total_debt_ratio
             if np.isfinite(capital_employed_def2) and capital_employed_def2 > 1.0:
                 capital_employed = capital_employed_def2; ce_source = "E+D"
    if not np.isfinite(capital_employed):
         ce_source = "Invalid/Unavailable"; ratio_log.append("⚠️ Capital Employed could not be reliably calculated. ROCE affected.")

    # --- Calculate Ratios (Result is NaN if components invalid/zero denominator) ---
    ratios['P/E'] = (current_price_val / eps) if valid_price and np.isfinite(eps) and eps != 0 else np.nan
    ratios['P/B'] = (current_price_val / bvps) if valid_price and np.isfinite(bvps) and bvps != 0 else np.nan
    ratios['EV/EBITDA'] = (enterprise_val / ebitda) if valid_enterprise_val and np.isfinite(ebitda) and ebitda != 0 else np.nan
    ratios['ROE'] = (net_income / total_equity) if np.isfinite(net_income) and np.isfinite(total_equity) and total_equity != 0 else np.nan
    ratios['ROA'] = (net_income / total_assets) if np.isfinite(net_income) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['ROCE'] = (ebit / capital_employed) if np.isfinite(ebit) and np.isfinite(capital_employed) and capital_employed != 0 else np.nan
    ratios['Asset Turnover'] = (revenue / total_assets) if np.isfinite(revenue) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['Current Ratio'] = (current_assets / current_liab) if np.isfinite(current_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan
    quick_assets = (current_assets - inventory) if np.isfinite(current_assets) and np.isfinite(inventory) else np.nan
    ratios['Quick Ratio'] = (quick_assets / current_liab) if np.isfinite(quick_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan

    # --- Compare with User/Fallback Benchmarks & Prepare Data Tables ---
    display(Markdown("*(Comparing ratios to user/fallback benchmarks...)*"))
    for key in REQUIRED_BENCHMARK_RATIOS:
        company_value = ratios.get(key); is_percent_type = key in ['ROE', 'ROA', 'ROCE']
        value_fmt = format_percent(company_value, 1) if is_percent_type else format_currency(company_value, '', 2, is_indian=False)
        if value_fmt == "N/A" and company_value is not None and not math.isfinite(company_value): value_fmt = "Inf"

        user_benchmark = user_benchmark_dict.get(key, None); fallback_benchmark = DEFAULT_BENCHMARK_RATIOS.get(key, None)
        benchmark_source = benchmark_source_dict.get(key, "?"); final_benchmark = None
        if benchmark_source == 'User Input' and user_benchmark is not None: final_benchmark = user_benchmark
        elif benchmark_source == 'Fallback Default' and fallback_benchmark is not None: final_benchmark = fallback_benchmark
        else: benchmark_source = "Not Available"

        bench_fmt = "N/A"
        if final_benchmark is not None: bench_fmt = f"{final_benchmark:.1f}%" if is_percent_type else f"{final_benchmark:.2f}"; bench_fmt += f" <i style='font-size:0.8em; color:grey'>({benchmark_source.replace('Fallback ','')})</i>"
        else: bench_fmt = "<i style='color:grey'>Not Provided</i>"

        status = "<i style='color:grey'>N/A</i>"; status_color = "grey"
        if np.isfinite(company_value) and final_benchmark is not None and np.isfinite(final_benchmark):
            numeric_value = company_value; numeric_benchmark = final_benchmark
            good_color = '#2ecc71'; bad_color = '#e74c3c'; neutral_color = '#f39c12'; grey_color='#7f8c8d'
            if abs(numeric_benchmark) > 1e-9: ratio_diff = (numeric_value - numeric_benchmark) / abs(numeric_benchmark)
            else: ratio_diff = float('inf') if numeric_value > 0 else float('-inf') if numeric_value < 0 else 0.0
            if key in ['ROE', 'ROA', 'ROCE', 'Asset Turnover', 'Current Ratio', 'Quick Ratio']: # Higher is better
                 if ratio_diff > 0.20: status_color = good_color; status = f"⬆️ Strong"
                 elif ratio_diff < -0.20: status_color = bad_color; status = f"⬇️ Weak"
                 else: status_color = grey_color; status = f"~ Average"
            elif key in ['P/E', 'P/B', 'EV/EBITDA']: # Lower might be better (valuation) or neutral
                 if ratio_diff < -0.20: status_color = good_color; status = f"✅ Low"
                 elif ratio_diff > 0.20: status_color = neutral_color; status = f"⚠️ High"
                 else: status_color = grey_color; status = f"~ Average"
            status = f"<span style='color:{status_color}'>{status}</span>"
            plot_value = numeric_value * 100 if is_percent_type else numeric_value; plot_benchmark = numeric_benchmark
            ratio_plotly_data['Ratio'].append(key); ratio_plotly_data['Company'].append(plot_value); ratio_plotly_data['Benchmark'].append(plot_benchmark);
            ratio_plotly_data['Benchmark Source'].append(benchmark_source); ratio_plotly_data['Status'].append(status)
        elif not np.isfinite(company_value): status = "<i style='color:grey'>Company N/A</i>"
        elif final_benchmark is None: status = "<i style='color:grey'>No Benchmark</i>"
        ratio_data_for_table.append([f"<b>{key}</b>", value_fmt, bench_fmt, status])

    # --- Display Logs & Table ---
    if ratio_log: display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
    if ratio_data_for_table:
        header = ["Ratio", f"{company_nm} Value", "Benchmark (Source)", "Status"]; html_table = "<p>(*Status vs User/Default benchmarks. Threshold: +/- 20%.)</p>"; html_table += "<table border='1' style='border-collapse:collapse; width:98%; margin-left:auto; margin-right:auto; font-size: 0.9em;'>"; html_table += "<thead style='background-color:#e9ecef; color: #495057;'><tr>" + "".join(f"<th style='padding: 8px; text-align:center;'>{h}</th>" for h in header) + "</tr></thead>"; html_table += "<tbody>"
        for i, row in enumerate(ratio_data_for_table): bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff"; html_table += f"<tr style='background-color:{bg_color};'>"; html_table += f"<td style='padding: 6px; text-align:left;'>{row[0]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[1]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[2]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[3]}</td>"; html_table += "</tr>"
        html_table += "</tbody></table><br>"; display(HTML(html_table))
    else: display(Markdown("*No ratio data available.*"))

    # --- Display Ratio Comparison Chart (Plotly Bullet) ---
    if ratio_plotly_data['Ratio']:
       num_ratios = len(ratio_plotly_data['Ratio']); fig_ratios = go.Figure()
       def get_range_colors(k): return ["#fadbd8", "#fdebd0", "#d5f5e3"] if k in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio'] else ["#d5f5e3","#fdebd0","#fadbd8"] if k in ['P/E','P/B','EV/EBITDA'] else ['lightgrey','darkgrey','grey']
       for i in range(num_ratios):
           ratio_name = ratio_plotly_data['Ratio'][i]; company_val = ratio_plotly_data['Company'][i]; benchmark_val = ratio_plotly_data['Benchmark'][i]; bench_src_plot = ratio_plotly_data['Benchmark Source'][i]
           if not np.isfinite(company_val) or not np.isfinite(benchmark_val): print(f"Skipping plot for {ratio_name}: Invalid Value ({company_val}) or Bench ({benchmark_val})."); continue
           range_colors = get_range_colors(ratio_name); is_percent = ratio_name in ['ROE', 'ROA', 'ROCE']
           poor_f=0.8; good_f=1.2; range1_limit = benchmark_val * (poor_f if benchmark_val >=0 else good_f); range2_limit = benchmark_val * (good_f if benchmark_val >=0 else poor_f)
           abs_max = max(abs(company_val), abs(benchmark_val)); plot_max_val = abs_max * 1.7 + 1; plot_min_val = 0
           if company_val < 0 or benchmark_val < 0: plot_min_val = min(company_val, benchmark_val) * 1.7 - 1
           if ratio_name in ['P/E', 'EV/EBITDA']: plot_max_val = min(plot_max_val, 100); plot_min_val = max(plot_min_val, -20)
           elif ratio_name in ['P/B']: plot_max_val = min(plot_max_val, 20); plot_min_val = max(plot_min_val, -5)
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: plot_max_val = min(plot_max_val, 5); plot_min_val = max(plot_min_val, 0)
           elif is_percent: plot_max_val = min(plot_max_val, 75); plot_min_val= max(plot_min_val,-50)
           else: plot_max_val = min(plot_max_val, 1000)
           if plot_min_val >= plot_max_val: plot_max_val = plot_min_val + 1
           steps = [];
           if abs(benchmark_val) < 1e-9: steps = [{'range': [plot_min_val, plot_max_val], 'color':'lightgrey'}]
           else: step_ranges = sorted([plot_min_val, range1_limit, range2_limit, plot_max_val]); step_colors = range_colors if benchmark_val >= 0 else range_colors[::-1]; steps = [{'range': [step_ranges[j], step_ranges[j+1]], 'color': step_colors[j]} for j in range(3)]
           bench_fmt_title = f"{benchmark_val:.1f}%" if is_percent else f"{benchmark_val:.2f}"; title_text = f"<b>{ratio_name}</b><br><i style='font-size:0.8em'>(Bench: {bench_fmt_title} [{bench_src_plot.replace('Fallback ','')}])</i>"
           lower_is_better = ratio_name in ['P/E', 'P/B', 'EV/EBITDA']; delta_decreasing_color = good_color if lower_is_better else bad_color; delta_increasing_color = bad_color if lower_is_better else good_color
           fig_ratios.add_trace(go.Indicator(
                mode = "number+gauge+delta", value = company_val, number={'suffix': "%" if is_percent else "", 'valueformat': ".1f"},
                delta = {'reference': benchmark_val, 'relative': True, 'valueformat':".0%", 'decreasing': {'color': delta_decreasing_color}, 'increasing': {'color': delta_increasing_color}},
                domain = {'row': i, 'column': 0}, title = {'text' : title_text, 'font':{'size':11}},
                gauge = { 'shape': "bullet", 'axis': {'range': [plot_min_val, plot_max_val] if plot_min_val < plot_max_val else [plot_min_val, plot_min_val+1], 'tickformat':'.1f'},
                    'threshold': { 'line': {'color': "black", 'width': 2}, 'thickness': 0.8, 'value': benchmark_val },
                    'steps': [ s for s in steps if s and isinstance(s.get('range'), list) and len(s['range'])==2 and s['range'][0] < s['range'][1] ],
                    'bar': {'color': "#2c3e50", 'thickness': 0.5} }
            ))
       fig_ratios.update_layout(height = max(300, num_ratios * 85), grid = {'rows': num_ratios, 'columns': 1, 'pattern': "independent"}, title_text='Ratio Performance vs. Benchmarks', title_x=0.5, margin=dict(l=160, r=40, t=60, b=30))
       fig_ratios.show()
    else: display(Markdown("*No valid ratios/benchmarks for comparison chart.*"))

    return ratios, ratio_data_for_table

# --- Execute Ratio Analysis ---
# Initialize ratios_calculated and ratio_table before the try block
ratios_calculated = {}; ratio_table = []

# Check if final_valuation and wacc_components are dictionaries and final_valuation has no error
if isinstance(final_valuation, dict) and isinstance(wacc_components, dict) and not final_valuation.get('error'):
    try:
        # Retrieve enterprise value safely from final_valuation dict
        ev_for_ratios = final_valuation.get('enterprise_value', np.nan)
        ratios_calculated, ratio_table = analyze_ratios(
            stock_data['info'], stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'],
            ev_for_ratios, current_price, shares_outstanding, currency_symbol,
            user_benchmarks, benchmark_sources, wacc_components, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Ratio Analysis: {e}**"))
         # Ensure they are reset on error
         ratios_calculated = {}; ratio_table = []
else:
    display(Markdown("❌ **Skipping Ratio Analysis due to prior errors or invalid data structures.**"));
    # Ensure they are reset if skipped
    ratios_calculated = {}; ratio_table = []


# %% [markdown]
# ## 9. Summary & Visualization Dashboard

# %%
# Dashboard Function Definition (No changes here, relies on valid inputs)
def display_summary_dashboard(company_nm, ticker_sym, current_pr, intrinsic_val_per_sh,
                              wacc_val, term_growth, init_growth, growth_src,
                              final_val_dict, hist_df, fcff_res, currency_sym, info_dict, input_src_dict,
                              is_indian): # Added flag

    display(Markdown(f"## Valuation Summary Dashboard: {company_nm} ({ticker_sym})"))
    source_summary = " | ".join(f"{k}:`{v.split('(')[0].strip()}`" for k, v in input_src_dict.items())
    growth_src_display = growth_src if growth_src else 'N/A'
    display(Markdown(f"**Input Sources:** {source_summary} | Growth:`{growth_src_display}`"))

    valid_intrinsic = intrinsic_val_per_sh is not None and np.isfinite(intrinsic_val_per_sh)
    valid_current_pr = current_pr is not None and np.isfinite(current_pr) and current_pr > 0
    valid_wacc = wacc_val is not None and np.isfinite(wacc_val)
    valid_term_g = term_growth is not None and np.isfinite(term_growth)
    valid_init_g = init_growth is not None and np.isfinite(init_growth)

    # --- Indicator Subplots ---
    fig_indicators = make_subplots(rows=1, cols=3, specs=[[{'type': 'indicator'}, {'type': 'indicator'}, {'type': 'indicator'}]],
                                 subplot_titles=("Price vs. Intrinsic", "WACC", "Growth Rates"), horizontal_spacing=0.1, column_widths=[0.4, 0.3, 0.3])

    # Indicator 1: Price vs Intrinsic
    intrinsic_val_num = intrinsic_val_per_sh if valid_intrinsic else np.nan
    intrinsic_val_fmt = format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian) if valid_intrinsic else "N/A"
    current_pr_num = current_pr if valid_current_pr else 0.0; upside_fmt = ""; delta_config = None
    if valid_intrinsic and valid_current_pr :
        if intrinsic_val_num > 0: upside_pct = (intrinsic_val_num / current_pr_num) - 1; upside_fmt = f" ({upside_pct:+.1%})"
        comparison_text = f"Intrinsic: {intrinsic_val_fmt}{upside_fmt}"
        delta_config = {'reference': intrinsic_val_num, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym, 'decreasing': {'color': '#2ecc71', 'symbol': '▲ Upside'}, 'increasing': {'color': '#e74c3c', 'symbol': '▼ Downside'}}
    elif valid_intrinsic: comparison_text = f"Intrinsic: {intrinsic_val_fmt} (Current N/A)"
    else: comparison_text = f"Intrinsic: N/A"
    fig_indicators.add_trace(go.Indicator(mode = "number+delta" if delta_config else "number", value = current_pr_num if valid_current_pr else 0, number={'prefix': currency_sym + " " if is_indian else currency_sym, 'valueformat': '.2f', 'font':{'size':24}}, title = {"text": f"<span style='font-size:0.8em'>{comparison_text}</span>"}, delta = delta_config, domain = {'row': 0, 'column': 0}), row=1, col=1)

    # Indicator 2: WACC Gauge
    wacc_val_num = wacc_val * 100 if valid_wacc else None
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = wacc_val_num,
        number = {'suffix': "%", 'valueformat':'.1f' if valid_wacc else '', 'font':{'size':20}},
        gauge = {'axis': {'range': [MIN_WACC*100, MAX_WACC*100]}, 'bar': {'color': "darkblue" if valid_wacc else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc", 'steps' : [ {'range': [MIN_WACC*100, 9], 'color': "#d5f5e3"}, {'range': [9, 14], 'color': "#fdebd0"}, {'range': [14, MAX_WACC*100], 'color': "#fadbd8"}]} if valid_wacc else None,
        title = {"text": "<span style='font-size:0.8em'>WACC</span>"}, domain = {'row': 0, 'column': 1}), row=1, col=2)

    # Indicator 3: Growth Gauge
    term_growth_num = term_growth * 100 if valid_term_g else None; init_growth_fmt = format_percent(init_growth, 1) if valid_init_g else "N/A"
    growth_title_text = f"<span style='font-size:0.8em'>Terminal Growth (g)<br>(Initial: {init_growth_fmt})</span>"
    gauge_max_range = max(TERMINAL_GROWTH_RATE_CAP * 100 + 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) + 1)
    gauge_min_range = MIN_GROWTH * 100 - 1
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = term_growth_num,
         number = {'suffix': "%", 'valueformat':'.1f' if valid_term_g else '', 'font':{'size':20}},
         gauge = {'axis': {'range': [gauge_min_range, gauge_max_range], 'tickwidth': 1}, 'bar': {'color': "darkgrey" if valid_term_g else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [{'range': [gauge_min_range, 2.0], 'color': "lightcoral"}, {'range': [2.0, 4.5], 'color': "lightyellow"}, {'range': [4.5, gauge_max_range], 'color': "lightgreen"}],
                 'threshold' : {'line': {'color': "black", 'width': 3}, 'thickness': 0.75, 'value': term_growth_num } if valid_term_g else None } if valid_term_g else None,
         title = {"text": growth_title_text}, domain = {'row': 0, 'column': 2}), row=1, col=3)

    fig_indicators.update_layout(height=220, margin=dict(l=30, r=30, t=50, b=20)); fig_indicators.show()

    # --- Historical Price Chart ---
    if not hist_df.empty:
        # Use 'Close' price for the main history chart for broader compatibility
        price_col_hist = 'Close' if 'Close' in hist_df.columns else None
        if price_col_hist:
             fig_hist = px.line(hist_df, x=hist_df.index, y=price_col_hist, title=f'{company_nm} Price History vs. DCF'); fig_hist.update_layout(xaxis_title='Date', yaxis_title=f'Price ({currency_sym})', height=400, legend_title_text='Legend'); fig_hist.update_xaxes(rangeslider_visible=False)
             if valid_intrinsic: fig_hist.add_hline(y=intrinsic_val_num, line_dash="dash", line_color="#e74c3c", opacity=0.8, annotation_text=f"DCF Intrinsic: {format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian)}", annotation_position="bottom right", annotation_font_color="#e74c3c")
             if valid_current_pr: fig_hist.add_hline(y=current_pr_num, line_dash="dot", line_color="#3498db", opacity=0.8, annotation_text=f"Current: {format_currency(current_pr_num, currency_sym, is_indian=is_indian)}", annotation_position="top left", annotation_font_color="#3498db")
             fig_hist.show()
        else:
             display(Markdown("*Historical price data ('Close') unavailable for chart.*"))
    else: display(Markdown("*Historical price data unavailable.*"))

    # --- DCF EV Breakdown Pie Chart ---
    # Ensure fcff_res is a dictionary before accessing
    if isinstance(fcff_res, dict):
        pv_forecast_sum = fcff_res.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_res.get('pv_terminal_value', np.nan)
        valid_pvs = np.isfinite([pv_forecast_sum, pv_tv]); total_ev = (pv_forecast_sum + pv_tv) if all(valid_pvs) else np.nan
        if np.isfinite(total_ev) and total_ev != 0 :
            labels_pie = []; values_pie = []; pull_pie = []; colors = ['#4e79a7','#f28e2b']
            if abs(pv_forecast_sum) > 1e-6: labels_pie.append(f'PV({FORECAST_YEARS}yr FCFFs)'); values_pie.append(pv_forecast_sum); pull_pie.append(0)
            if abs(pv_tv) > 1e-6: labels_pie.append('PV(TV)'); values_pie.append(pv_tv); pull_pie.append(0.05 if len(labels_pie) > 1 else 0)
            if labels_pie:
                hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                if is_indian: hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                fig_dcf_pie = go.Figure(data=[go.Pie(labels=labels_pie, values=values_pie, hole=.4, pull=pull_pie, marker_colors=colors[:len(labels_pie)], hovertemplate=hover_format, textinfo='percent+label', insidetextorientation='radial', sort=False)])
                fig_dcf_pie.update_layout( title_text=f"DCF EV Breakdown (Total EV: {format_currency(total_ev, currency_sym, is_indian=is_indian)})", title_x=0.5, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), height=350, margin=dict(t=60, b=50, l=20, r=20) )
                fig_dcf_pie.show()
            else: display(Markdown(f"*DCF components zero/negligible. Cannot generate EV Breakdown chart.*"))
        else: display(Markdown(f"*EV calculation failed ({format_currency(total_ev, currency_sym, is_indian=is_indian)}). Cannot generate EV Breakdown chart.*"))
    else: display(Markdown(f"*FCFF results invalid. Cannot generate EV Breakdown chart.*"))

    # --- FCFF Projection Bar Chart ---
    # Ensure fcff_res is a dictionary before accessing
    if isinstance(fcff_res, dict) and fcff_res.get('projected_fcffs'):
        years = [f'Y{i+1}' for i in range(len(fcff_res['projected_fcffs']))]; pv_list_clean = [p if np.isfinite(p) else 0 for p in fcff_res.get('pv_fcffs', [])]; fcff_list_clean = [f if np.isfinite(f) else 0 for f in fcff_res.get('projected_fcffs', [])]
        if len(pv_list_clean) == len(years) and len(fcff_list_clean) == len(years):
            df_fcff_plot = pd.DataFrame({'Year': years, 'Forecasted FCFF': fcff_list_clean, 'PV of FCFF': pv_list_clean}); df_fcff_plot['Forecasted FCFF'] = pd.to_numeric(df_fcff_plot['Forecasted FCFF']); df_fcff_plot['PV of FCFF'] = pd.to_numeric(df_fcff_plot['PV of FCFF'])
            df_fcff_melt = df_fcff_plot.melt(id_vars='Year', var_name='Cash Flow Type', value_name='Value')
            fig_fcff_bar = px.bar(df_fcff_melt, x='Year', y='Value', color='Cash Flow Type', title='FCFF Forecast & Present Values', labels={'Value':f'Value ({currency_sym})'}, barmode='group', height=350, color_discrete_map={'Forecasted FCFF':'#5dade2', 'PV of FCFF':'#af7ac5'})
            hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            if is_indian: hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            fig_fcff_bar.update_traces(hovertemplate=hover_bar_format); fig_fcff_bar.update_layout(margin=dict(t=60, b=30, l=50, r=30)); fig_fcff_bar.show()
        else: display(Markdown("*FCFF projection data mismatch/invalid. Cannot plot.*"))
    else: display(Markdown("*FCFF projection data unavailable or invalid.*"))

    # --- Valuation Conclusion ---
    display(Markdown("### Valuation Conclusion"));
    # Ensure final_val_dict is a dictionary before accessing
    if isinstance(final_val_dict, dict):
        intrinsic_val = final_val_dict.get('intrinsic_value_per_share', np.nan)
        current_pr_num = current_pr if valid_current_pr else np.nan; conclusion = ""
        if final_val_dict.get('error'): conclusion = f"<span style='color:red'>❌ Valuation Error: {final_val_dict.get('error', '?')}</span>"
        elif not np.isfinite(intrinsic_val) or intrinsic_val <= 0: conclusion = f"⚠️ Valuation resulted in invalid/non-positive Intrinsic Value ({format_currency(intrinsic_val, currency_sym, is_indian=is_indian)})."
        elif not np.isfinite(current_pr_num): conclusion = f"✅ Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}. Cannot compare (Current Price N/A)."
        else:
             diff_pct = (intrinsic_val / current_pr_num) - 1
             current_fmt = format_currency(current_pr_num, currency_sym, is_indian=is_indian)
             intrinsic_fmt = format_currency(intrinsic_val, currency_sym, is_indian=is_indian)
             if abs(diff_pct) < 0.10: conclusion = f"<span style='color:grey'>~ Fairly Valued:</span> Price ({current_fmt}) vs DCF ({intrinsic_fmt}). Diff: {diff_pct:+.1%}"
             elif diff_pct < 0: overval_pct = (current_pr_num / intrinsic_val) - 1; conclusion = f"<span style='color:#e74c3c'>🔻 Potentially Overvalued:</span> Price ({current_fmt}) is **{overval_pct:+.1%}** above DCF ({intrinsic_fmt})."
             else: conclusion = f"<span style='color:#2ecc71'>✅ Potentially Undervalued:</span> Price ({current_fmt}) suggests **{diff_pct:+.1%}** upside to DCF ({intrinsic_fmt})."
    else:
        conclusion = f"<span style='color:red'>❌ Valuation Error: Final valuation data structure invalid.</span>"

    display(Markdown(f"> **{conclusion}**"))

# --- Execute Summary Dashboard Generation ---
# Check if results are valid dictionaries before passing them
if isinstance(final_valuation, dict) and isinstance(fcff_results, dict) and isinstance(stock_data.get('info'), dict) and isinstance(input_sources, dict) and not final_valuation.get('error'):
    try:
        display_summary_dashboard(
            company_name, ticker_symbol, current_price, final_valuation.get('intrinsic_value_per_share'),
            discount_rate, fcff_results.get('terminal_growth_rate'), fcff_results.get('initial_growth_rate'),
            fcff_results.get('growth_source', 'N/A'), final_valuation, stock_data['history'], fcff_results,
            currency_symbol, stock_data['info'], input_sources, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Dashboard Generation: {e}**"))
else:
    display(Markdown("❌ **Skipping Summary Dashboard due to prior errors or invalid data structures.**"))
    # Optionally print the error states for debugging:
    # print("Final Valuation State:", final_valuation)
    # print("FCFF Results State:", fcff_results)


# %% [markdown]
# ## 10. Assumptions & Disclaimer (Reflects Dynamic/User Inputs & Fallbacks - Growth Logic Updated)

# %%
# Dynamically generate assumption strings based on final values
# Add checks to ensure fcff_results and wacc_components are dictionaries before accessing
rf_source_disp = input_sources.get('Rf', '?')
rm_source_disp = input_sources.get('Rm','?')
beta_source_disp = input_sources.get('Beta','?')
c_spread_source_disp = input_sources.get('Credit Spread','N/A')
bench_source_disp = input_sources.get('Benchmarks','?')

# Safely get values from dictionaries, providing defaults if dicts don't exist or keys missing
growth_source_disp = fcff_results.get('growth_source', 'N/A') if isinstance(fcff_results, dict) else 'Error/Not Run'
tax_rate_source_disp = wacc_components.get('tax_rate_source', 'N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
equity_source_disp = wacc_components.get('equity_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
weights_source_disp = wacc_components.get('weights_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
terminal_growth_final = fcff_results.get('terminal_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan
initial_growth_final = fcff_results.get('initial_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan


# %%
# ----> UPDATED TEXT TO REFLECT NEW GROWTH LOGIC (V14) <----
display(Markdown(f"""
**Key Assumptions & Input Sources:**
*   **Financial Data:** Relies on data provided by Yahoo Finance (`yfinance`). Accuracy subject to provider. Missing data uses fallbacks or causes errors.
*   **Risk-Free Rate (Rf):** Attempted dynamic fetch, **confirmed/overridden by user.** (`Source: {rf_source_disp}` - Default suggestion: `{DEFAULT_RF_RATE_SUGGESTION:.1f}%`)
*   **Expected Market Return (Rm):** Based on **user input.** (`Source: {rm_source_disp}`)
*   **Market Risk Premium (MRP):** Derived from user `Rm` and final `Rf`.
*   **Beta (β):** Dynamically fetched if valid, otherwise **required user input.** (`Source: {beta_source_disp}`)
*   **Cost of Debt (Kd):** Implied if possible & within bounds (`{format_percent(MIN_COST_OF_DEBT)}` - `{format_percent(MAX_COST_OF_DEBT)}`). Otherwise, defaults to `Rf + Credit Spread`.
*   **Credit Spread:** Used only if default Kd triggered. Based on **user input** (default suggestion: `{format_percent(DEFAULT_CREDIT_SPREAD)}`). (`Source: {c_spread_source_disp}`)
*   **Growth Rates:**
    *   **Initial Growth:** `{format_percent(initial_growth_final)}` used for Year 1, determined via: (`Source: {growth_source_disp}`)
        1.  Fetch last 3 years (4 data points) of Revenue, FCFF, and Stock Price (daily).
        2.  **Validate Data:** Check if T-0 (end) is finite and T-3 (start) is positive and finite for each metric.
        3.  Calculate **3-Year CAGR** *only* for metrics with valid start/end data (using `calculate_cagr_modified`). Stock CAGR uses daily price history, preferring 'Adj Close', falling back to 'Close'.
        4.  Calculate the **Geometric Mean** (`geometric_mean_modified`) of the successfully calculated CAGRs from step 3.
        5.  If GM calculation fails or *no valid CAGRs were computed* in step 3, use **Default (`{format_percent(DEFAULT_GROWTH_RATE)}`)**.
        6.  Apply **Bounds:** The result from step 4 or 5 is capped between **`{format_percent(MIN_INITIAL_GROWTH_FLOOR)}` (Floor)** and **`{format_percent(MAX_INITIAL_GROWTH_CEILING)}` (Ceiling)**.
    *   **Terminal Growth (g):** `{format_percent(terminal_growth_final)}` = min(GDP cap=`{format_percent(TERMINAL_GROWTH_RATE_CAP)}`, WACC*`0.67`, Final Initial Growth, Rf*`0.95`), floored at `{format_percent(MIN_GROWTH)}`, ensured `< WACC`.
*   **Tax Rate:** Effective rate `{tax_rate_source_disp}` (bounded `{format_percent(MIN_TAX_RATE)}` - `{format_percent(MAX_TAX_RATE)}`) or default (`{format_percent(DEFAULT_TAX_RATE)}`).
*   **Capital Structure:** Assumed stable using `{equity_source_disp}` and calculated debt. (`Weights Source: {weights_source_disp}`)
*   **Industry Benchmarks:** Uses **user input** or **generic fallbacks** if input left blank. (`Source: {bench_source_disp}`)
*   **Currency Formatting:** Uses ₹/Lakh/Crore for `.NS` tickers, standard K/M/B/T otherwise.

**Limitations:**
*   **Model Sensitivity:** DCF highly sensitive to inputs. **Sensitivity analysis (not included) crucial.**
*   **Input Quality:** Validity depends on user inputs and appropriateness of fallbacks.
*   **Historical Data Availability:** Financial statement data (needed for Rev/FCFF CAGR) is often limited to ~4 years via `yfinance`. Accuracy of 3-year CAGR depends heavily on these few points. Stock price history is usually more extensive but relies on finding suitable start/end dates. CAGR calculation is skipped if start/end data is invalid or start is non-positive.
*   **FCFF Calculation & Projection:** FCFF calculation itself can be sensitive to accounting choices (e.g., CapEx, WC changes). If the **latest historical FCFF (T-0) cannot be calculated**, the model **cannot project FCFF** and the DCF valuation will fail, even if a growth rate was estimated from other sources.
*   **CAGR/GM Modifications:** The modified CAGR handles non-positive end values (relative to positive start) but requires a positive start value. The modified GM handles negative rates but falls back to arithmetic mean if rates are <= -100% or other errors occur. GM is less meaningful with highly disparate or negative rates.
*   **Dynamic Data Reliability:** Depends on `yfinance` accuracy/availability. Rf proxies need verification. Stock history column names ('Adj Close', 'Close') assumed.
*   **Simplifications:** Linear decay, constant rates/structure. Ignores qualitative factors, macro shocks, sentiment.
*   **Fallback Accuracy:** Default growth rate (6.5%) and generic benchmarks may not be representative.
*   **Data Completeness:** Performance degrades with missing data. *Cross-verify key data.*

**Disclaimer:**
*For **educational and illustrative purposes only**. **NOT investment advice**. Models have limitations. Users MUST conduct own due diligence and consult a qualified financial advisor before investing. Creators NOT liable for actions based on this report.*
"""))



# %%

````

````python
# %% [markdown]
# # Dynamic Stock Financial Analysis & DCF Valuation
# **(Interactive Report - Dynamic/User Input Driven Assumptions - V21 - Detailed Avg FCFF Logging)**
#
# **Analyst:** Simulated AI Research Analyst
#
# **Objective:** To perform a comprehensive financial analysis and estimate the intrinsic value per share using DCF. Prioritizes dynamic data, user input, and defaults. **FCFF projection starts from a *positive average* FCFF determined via a fallback sequence, with detailed logging of each method attempt, yearly values, and average. Initial growth still uses GM of individual 3yr CAGRs.**
#
# **Methodology:**
# 1.  **User Input & Dynamic Fetching:** Standard setup.
# 2.  **Data Acquisition:** Standard setup, including 4y daily history.
# 3.  **WACC Calculation:** Standard setup.
# 4.  **FCFF Calculation & Projection:**
#     *   Calculate historical FCFF (T-0 to T-3) individually using standard fallbacks (`_calculate_fcff_single_year_best_estimate`) for CAGR calculation.
#     *   **Determine Projection Basis (`fcff_start_basis`):** Use new helper `get_positive_average_fcff_basis_v21` with strict positive-average fallback sequence. **Logs detailed attempts for each averaging method.**
#     *   **Halt if No Positive Basis:** If `fcff_start_basis` is NaN, set error and stop FCFF projection.
#     *   **Estimate initial growth:** (Uses individual historical FCFFs for FCFF CAGR) Calculate 3yr CAGRs for Rev, FCFF, Stock. Calculate GM. Fallback/Bound.
#     *   **Project FCFF (Conditional):** Project forward from `fcff_start_basis` using decaying growth rate.
# 5.  **Terminal Value:** Standard calculation.
# 6.  **Discounting & Valuation:** Standard calculation.
# 7.  **Ratio Analysis:** Standard calculation.
# 8.  **Interactive Visualization:** Standard setup.

# %% [markdown]
# ## 1. Setup and Libraries

# %%
# Standard Libraries
import warnings
import sys
import numpy as np
import pandas as pd
import datetime
import math # For isnan, isinf checks

# Data Fetching
import yfinance as yf

# Visualization (Using Plotly)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

# Display Utilities
from IPython.display import display, Markdown, HTML

# Settings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pio.templates.default = "plotly_white" # Cleaner Plotly template

# %% [markdown]
# ## 2. Configuration and Helper Functions (FCFF Helper Updated V21)

# %%
# --- Model Configuration ---
DEFAULT_TICKER = "MUTHOOTFIN.NS"
FORECAST_YEARS = 5
TERMINAL_GROWTH_RATE_CAP = 0.05
DEFAULT_GROWTH_RATE = 0.065
DEFAULT_RF_RATE_SUGGESTION = 7.0
DEFAULT_CURRENCY_SYMBOL = "₹"
DEFAULT_TAX_RATE = 0.25
DEFAULT_CREDIT_SPREAD = 0.015
DEFAULT_MARKET_RISK_PREMIUM = 0.06

# --- Generic Fallback Benchmark Ratios ---
DEFAULT_BENCHMARK_RATIOS = {
    'P/E': 17.3, 'P/B': 2.5, 'EV/EBITDA': 11.0, 'ROE': 19.05,
    'ROA': 3.92, 'ROCE': 11.85,'Asset Turnover': 0.15,
    'Current Ratio': 1.5, 'Quick Ratio': 1.5
}

# --- Model Stability Bounds ---
MIN_WACC = 0.06; MAX_WACC = 0.25; MIN_TAX_RATE = 0.10; MAX_TAX_RATE = 0.40
MIN_GROWTH = 0.00; MIN_INITIAL_GROWTH_FLOOR = 0.011; MAX_INITIAL_GROWTH_CEILING = 0.15
MIN_COST_OF_DEBT = 0.03; MAX_COST_OF_DEBT = 0.18

# --- Ratio Keys Needed for Benchmarking ---
REQUIRED_BENCHMARK_RATIOS = list(DEFAULT_BENCHMARK_RATIOS.keys())

# --- Helper Functions --- (safe_get, formatters, input validation - unchanged)
def print_keys(statement, name):
    print(f"\n--- Available Keys in {name} ---")
    if isinstance(statement, pd.DataFrame):
        valid_indices = [idx for idx in statement.index if idx is not None]
        print(", ".join(map(str, valid_indices)))
    elif isinstance(statement, dict): print(", ".join(statement.keys()))
    print("------------------------------------\n")

def safe_get(data, key_list, statement_name="Statement", default=np.nan): # Default to NaN
    if not isinstance(key_list, list): key_list = [key_list]
    value = None
    for key in key_list:
      current_value = None; raw_val = None
      if isinstance(data, pd.Series):
          if key in data.index:
              try:
                  raw_val = data.loc[key]
                  if isinstance(raw_val, (int, float)) and np.isfinite(raw_val): current_value = float(raw_val)
                  elif isinstance(raw_val, pd.Series):
                       if not raw_val.empty and pd.notna(raw_val.iloc[0]) and np.isfinite(raw_val.iloc[0]): current_value = float(raw_val.iloc[0])
                  elif pd.notna(raw_val):
                      try_val = float(raw_val);
                      if np.isfinite(try_val): current_value = try_val
              except (KeyError, TypeError, IndexError, ValueError): pass
      elif isinstance(data, dict):
          raw_val = data.get(key)
          if raw_val is not None:
              try:
                  if isinstance(raw_val, str) and raw_val.lower() in ['n/a', 'none', '']: pass
                  else: temp_val = float(raw_val);
                  if np.isfinite(temp_val): current_value = temp_val
              except (TypeError, ValueError): pass
      if current_value is not None: value = current_value; break
    if value is None: value = default
    return value

def format_currency(value, symbol="₹", precision=2, is_indian=False):
    if value is None or pd.isna(value) or not math.isfinite(value): return "N/A"
    if abs(value) > 1e18: return "Inf"
    sign = "-" if value < 0 else ""; abs_val = abs(value)
    if is_indian:
        symbol = "₹"
        if abs_val >= 1e7: formatted_val = f"{abs_val / 1e7:.{precision}f} Cr"
        elif abs_val >= 1e5: formatted_val = f"{abs_val / 1e5:.{precision}f} Lakh"
        else: formatted_val = f"{abs_val:,.{precision}f}"
        return f"{sign}{symbol} {formatted_val}"
    else:
        if abs_val >= 1e12: formatted_val = f"{abs_val / 1e12:.{precision}f}T"
        elif abs_val >= 1e9: formatted_val = f"{abs_val / 1e9:.{precision}f}B"
        elif abs_val >= 1e6: formatted_val = f"{abs_val / 1e6:.{precision}f}M"
        elif abs_val >= 1e3: formatted_val = f"{abs_val / 1e3:.{precision}f}K"
        else: formatted_val = f"{abs_val:.{precision}f}"
        return f"{sign}{symbol}{formatted_val}"

def format_percent(value, precision=1):
     if value is None or pd.isna(value) or not math.isfinite(value): return "N/A"
     if abs(value) > 1e5: return "Inf"
     return f"{value * 100:.{precision}f}%"

def get_validated_float_input(prompt_message, default_value=None, allow_blank=False, min_val=None, max_val=None):
    while True:
        prompt_with_default = prompt_message
        if default_value is not None: prompt_with_default += f" [Default: {default_value}]: "
        elif allow_blank: prompt_with_default += f" [Leave blank to skip/use fallback]: "
        else: prompt_with_default += ": "
        try:
            user_input_str = input(prompt_with_default).strip()
            if not user_input_str:
                if allow_blank: return None
                if default_value is not None:
                    try: return float(default_value)
                    except ValueError: raise ValueError(f"Internal Error: Invalid default value '{default_value}'.")
                raise ValueError("Input cannot be empty.")
            if user_input_str.endswith('%'): user_input_str = user_input_str[:-1]
            value = float(user_input_str)
            if min_val is not None and value < min_val: raise ValueError(f"Value must be >= {min_val}.")
            if max_val is not None and value > max_val: raise ValueError(f"Value must be <= {max_val}.")
            return value
        except ValueError as e: print(f"❌ Invalid input: {e}. Please enter a valid number.")

def try_get_dynamic_rf(rf_ticker_symbol="^TNX"):
    print(f"Attempting to fetch dynamic Risk-Free Rate ({rf_ticker_symbol})...")
    try:
        rf_ticker = yf.Ticker(rf_ticker_symbol); hist = rf_ticker.history(period="5d")
        if not hist.empty and 'Close' in hist.columns:
            last_close = hist['Close'].iloc[-1]
            if pd.notna(last_close):
                dynamic_rf = last_close / 100.0
                if 0.001 < dynamic_rf < 0.20: print(f"✅ Dynamically fetched Rf ({rf_ticker_symbol}): {format_percent(dynamic_rf)}"); return dynamic_rf
                else: print(f"⚠️ Fetched Rf value ({format_percent(dynamic_rf)}) out of typical range. Ignoring."); return None
            else: print(f"⚠️ Fetched Rf data was invalid (NaN) for {rf_ticker_symbol}."); return None
        else: print(f"⚠️ Could not fetch recent data for {rf_ticker_symbol}."); return None
    except Exception as e: print(f"❌ Error fetching dynamic Rf ({rf_ticker_symbol}): {e}"); return None

# --- Growth Calculation Helpers (Unchanged from V18) ---
def calculate_cagr_modified(start_val, end_val, num_years):
    if pd.isna(start_val) or pd.isna(end_val) or not math.isfinite(start_val) or not math.isfinite(end_val) or num_years <= 0: return np.nan
    if start_val <= 1e-9: return np.nan
    cagr = np.nan
    try:
        if end_val <= 1e-9:
            end_val_abs = max(abs(end_val), 1e-9); ratio = end_val_abs / start_val
            if ratio > 0: cagr_magnitude = ratio ** (1 / num_years) - 1; cagr = -cagr_magnitude
        else:
            ratio = end_val / start_val
            if ratio > 0: cagr = ratio ** (1 / num_years) - 1
    except (ValueError, OverflowError, ZeroDivisionError) as e: cagr = np.nan
    if pd.notna(cagr) and not (-0.9999 < cagr < 10.0): cagr = np.nan
    return cagr

def geometric_mean_modified(rates):
    valid_rates = [r for r in rates if pd.notna(r) and math.isfinite(r)]
    if not valid_rates: return np.nan
    if len(valid_rates) == 1: return valid_rates[0]
    offset = 1.0; factors_offset = [r + offset for r in valid_rates]
    if any(f <= 1e-9 for f in factors_offset):
         print("⚠️ Geometric mean calculation failed due to rate <= -100%. Falling back to arithmetic mean.")
         return np.mean(valid_rates)
    n = len(valid_rates); product = 1.0
    try:
        for factor in factors_offset: product *= factor
        if product < 0 and n % 2 == 0:
             print("⚠️ Geometric mean calculation failed (negative base for even root). Falling back to arithmetic mean.")
             return np.mean(valid_rates)
        gm_offset = np.sign(product) * (np.abs(product) ** (1/n)); gm = gm_offset - offset
        if pd.notna(gm) and not (-0.9999 < gm < 10.0):
            print(f"⚠️ Geometric mean result ({format_percent(gm)}) out of reasonable bounds. Falling back to arithmetic mean.")
            return np.mean(valid_rates)
        return gm
    except (ValueError, OverflowError, ZeroDivisionError, TypeError) as e:
        print(f"⚠️ Geometric mean calculation encountered numeric error ({e}). Falling back to arithmetic mean.")
        return np.mean(valid_rates)

# --- Internal FCFF Component Fetcher (Unchanged from V18) ---
def _get_fcff_components(fin_col, bs_col, cf_col, bs_prev_col):
    """Fetches all components needed for various FCFF calculations for a single year."""
    components = {}
    components['ebit'] = safe_get(fin_col, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
    components['dep_amort'] = safe_get(cf_col, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=np.nan)
    capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
    components['capex'] = abs(safe_get(cf_col, capex_keys, default=np.nan))
    wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]
    components['wc_cf'] = safe_get(cf_col, wc_keys, default=np.nan)
    cfo_keys = ["Cash Flow From Operating Activities", "Operating Cash Flow", "Total Cash From Operating Activities"]
    components['cfo'] = safe_get(cf_col, cfo_keys, default=np.nan)
    interest_expense_keys = ['Interest Expense', 'Interest Expense Net']
    components['int_exp'] = abs(safe_get(fin_col, interest_expense_keys, default=np.nan))
    components['tax_prov'] = safe_get(fin_col, ['Tax Provision', 'Income Tax Expense'], default=np.nan)
    components['revenue'] = safe_get(fin_col, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    if np.isfinite(components['ebit']) and np.isfinite(components['dep_amort']):
        components['ebitda'] = components['ebit'] + components['dep_amort']
    else: components['ebitda'] = np.nan
    components['wc'] = components['wc_cf']; components['wc_src'] = "CF"
    if not np.isfinite(components['wc']):
        components['wc_src'] = "BS Calc"
        if isinstance(bs_prev_col, pd.Series) and isinstance(bs_col, pd.Series):
             curr_assets = safe_get(bs_col, ['Current Assets', 'Total Current Assets'], default=np.nan); curr_liab = safe_get(bs_col, ['Current Liabilities', 'Total Current Liabilities'], default=np.nan)
             curr_wc_val = curr_assets - curr_liab if np.isfinite(curr_assets) and np.isfinite(curr_liab) else np.nan
             prev_assets = safe_get(bs_prev_col, ['Current Assets', 'Total Current Assets'], default=np.nan); prev_liab = safe_get(bs_prev_col, ['Current Liabilities', 'Total Current Liabilities'], default=np.nan)
             prev_wc_val = prev_assets - prev_liab if np.isfinite(prev_assets) and np.isfinite(prev_liab) else np.nan
             if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val): components['wc'] = curr_wc_val - prev_wc_val
             else: components['wc'] = np.nan; components['wc_src'] = "BS Fail"
        else: components['wc'] = np.nan; components['wc_src'] = "No Prev BS"
    # Apply absolute value after check
    if np.isfinite(components['capex']): components['capex'] = abs(components['capex'])
    if np.isfinite(components['int_exp']): components['int_exp'] = abs(components['int_exp'])
    return components

# --- FCFF Calculation Helper: Individual Year Best Estimate (for HISTORICAL list for CAGR) V19 ---
def _calculate_fcff_single_year_best_estimate(fin_col, bs_col, cf_col, bs_prev_col, tax_rt):
    """
    Calculates single year FCFF using NOPAT -> CFO+IntAdj -> CFO-Capex fallbacks.
    Returns best possible finite value (can be negative) or NaN. Method string also returned.
    """
    c = _get_fcff_components(fin_col, bs_col, cf_col, bs_prev_col)
    if not np.isfinite(tax_rt): tax_rt = np.nan
    # Method 1: NOPAT
    if all(map(np.isfinite, [c['ebit'], tax_rt, c['dep_amort'], c['capex'], c['wc']])):
        nopat_m1 = c['ebit'] * (1 - tax_rt); fcff_m1 = nopat_m1 + c['dep_amort'] - c['capex'] - c['wc']
        if np.isfinite(fcff_m1): return fcff_m1, f"NOPAT (WC:{c['wc_src']})"
    # Method 2: CFO+IntAdj
    if all(map(np.isfinite, [c['cfo'], c['capex'], c['int_exp'], tax_rt])):
        int_tax_shield = c['int_exp'] * (1 - tax_rt); fcff_m2 = c['cfo'] - c['capex'] + int_tax_shield
        if np.isfinite(fcff_m2): return fcff_m2, "CFO+IntAdj"
    # Method 3: CFO-CapEx
    if all(map(np.isfinite, [c['cfo'], c['capex']])):
        fcff_m3 = c['cfo'] - c['capex']
        if np.isfinite(fcff_m3): return fcff_m3, "CFO-CapEx"
    return np.nan, "All Methods Failed"


# --- FCFF T-0 Calculation Helper with Positive Average Priority (V21 - Detailed Logging) ---
def get_positive_average_fcff_basis_v21(financials_df, balance_sheet_df, cashflow_df, tax_rt, log_list):
    """
    Determines the T-0 FCFF starting value for projection using a strict positive-average priority sequence
    with detailed logging of each step.
    Returns (positive_avg_fcff, method_name) or (np.nan, "No Positive Average FCFF Found").
    """
    max_avg_years = 3 # Use last 3 years for averaging: T-0, T-1, T-2
    num_statement_years = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    log_list.append("--- Determining T-0 Projection Basis (Positive Average Priority V21) ---")

    if num_statement_years < max_avg_years:
         log_list.append(f"  ❌ Insufficient data for {max_avg_years}-year averaging ({num_statement_years} years found).")
         # Final fallback: calculate T-0 individually using best estimate
         fin_t0 = financials_df.iloc[:, 0]; cf_t0 = cashflow_df.iloc[:, 0]
         bs_t0 = balance_sheet_df.iloc[:, 0] if not balance_sheet_df.empty else None
         bs_t1 = balance_sheet_df.iloc[:, 1] if balance_sheet_df.shape[1] > 1 else None
         fcff_t0_ind, method_t0_ind = _calculate_fcff_single_year_best_estimate(fin_t0, bs_t0, cf_t0, bs_t1, tax_rt)
         if np.isfinite(fcff_t0_ind) and fcff_t0_ind > 1e-9:
             log_list.append(f"  ✅ Using T-0 Individual Best Estimate as basis (positive): {format_currency(fcff_t0_ind)} (Method: {method_t0_ind})")
             return fcff_t0_ind, f"{method_t0_ind} (T-0 Only/Insufficient Avg Data)"
         else:
             log_list.append(f"  ❌ T-0 Individual Best Estimate is non-positive/invalid: {format_currency(fcff_t0_ind)} (Method: {method_t0_ind}). No positive basis found.")
             return np.nan, "No Positive T-0 FCFF Found (Insufficient Avg Data)"

    # --- Define Calculation Functions for Each Method (Single Year) ---
    def calc_method0(fin, bs, cf, bs_prev): # EBITDA - TaxProv - WC - CapEx
        c = _get_fcff_components(fin, bs, cf, bs_prev)
        if all(map(np.isfinite, [c['ebitda'], c['tax_prov'], c['wc'], c['capex']])):
            return c['ebitda'] - c['tax_prov'] - c['wc'] - c['capex']
        return np.nan

    def calc_method1(fin, bs, cf, bs_prev, tax_rate): # NOPAT
        c = _get_fcff_components(fin, bs, cf, bs_prev)
        if all(map(np.isfinite, [c['ebit'], tax_rate, c['dep_amort'], c['capex'], c['wc']])):
            nopat = c['ebit'] * (1 - tax_rate)
            return nopat + c['dep_amort'] - c['capex'] - c['wc']
        return np.nan

    def calc_method2(fin, bs, cf, bs_prev, tax_rate): # CFO + IntAdj
        c = _get_fcff_components(fin, bs, cf, bs_prev)
        if all(map(np.isfinite, [c['cfo'], c['capex'], c['int_exp'], tax_rate])):
            int_tax_shield = c['int_exp'] * (1 - tax_rate)
            return c['cfo'] - c['capex'] + int_tax_shield
        return np.nan

    def calc_method3(fin, bs, cf, bs_prev): # CFO - CapEx
        c = _get_fcff_components(fin, bs, cf, bs_prev)
        if all(map(np.isfinite, [c['cfo'], c['capex']])):
            return c['cfo'] - c['capex']
        return np.nan

    def calc_method4(fin, bs, cf, bs_prev): # EBITDA - TaxProv (Proxy for Method 5 Avg)
        c = _get_fcff_components(fin, bs, cf, bs_prev)
        if all(map(np.isfinite, [c['ebitda'], c['tax_prov']])):
            return c['ebitda'] - c['tax_prov']
        return np.nan

    def calc_method5(fin, bs, cf, bs_prev): # EBITDA Margin (Proxy for Method 6 Avg)
        c = _get_fcff_components(fin, bs, cf, bs_prev)
        if all(map(np.isfinite, [c['ebitda'], c['revenue']])) and c['revenue'] > 1e-9:
            return c['ebitda'] / c['revenue']
        return np.nan

    # --- Try Averaging Methods Sequentially with Detailed Logging ---
    methods_to_try_avg = [
        ("EBITDA-TaxProv-WC-Capex Avg", calc_method0, False), # (User Priority)
        ("NOPAT Avg", calc_method1, True),
        ("CFO+IntAdj Avg", calc_method2, True),
        ("CFO-CapEx Avg", calc_method3, False)
    ]

    for method_name, calc_func, needs_tax_rate in methods_to_try_avg:
        log_list.append(f"  Trying Method: {method_name}...")
        yearly_fcffs = []
        valid_count = 0
        log_details = []
        for i in range(max_avg_years):
             fin = financials_df.iloc[:, i]; bs = balance_sheet_df.iloc[:, i]; cf = cashflow_df.iloc[:, i]
             bs_prev = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None
             tax_arg = tax_rt if needs_tax_rate else () # Pass tax only if needed
             # Handle args correctly for functions needing/not needing tax_rt
             if needs_tax_rate:
                 fcff_val = calc_func(fin, bs, cf, bs_prev, tax_arg)
             else:
                 fcff_val = calc_func(fin, bs, cf, bs_prev)

             yearly_fcffs.append(fcff_val)
             log_details.append(f"T-{i}: {format_currency(fcff_val, '', 0)}")
             if np.isfinite(fcff_val): valid_count += 1

        log_list.append(f"    Values: [{', '.join(log_details)}]")
        if valid_count > 0:
            avg_fcff = np.nanmean(yearly_fcffs) # nanmean ignores NaNs
            log_list.append(f"    Average ({valid_count} valid yrs): {format_currency(avg_fcff, '', 0)}")
            if np.isfinite(avg_fcff) and avg_fcff > 1e-9:
                log_list.append(f"  ✅ Found Positive Average. Using as projection basis.")
                return avg_fcff, method_name
            else:
                log_list.append(f"    Average is non-positive or invalid. Trying next method.")
        else:
            log_list.append(f"    No valid values for this method in the last {max_avg_years} years. Trying next method.")

    # --- Try Fallback Averages if standard methods failed ---
    # Method 4 Avg: Avg(EBITDA - TaxProv)
    log_list.append(f"  Trying Method: Avg(EBITDA-TaxProv)...")
    yearly_diffs = []
    valid_count = 0
    log_details = []
    for i in range(max_avg_years):
        fin = financials_df.iloc[:, i]; cf = cashflow_df.iloc[:, i]
        diff_val = calc_method4(fin, None, cf, None) # Method 4 needs fin/cf
        yearly_diffs.append(diff_val)
        log_details.append(f"T-{i}: {format_currency(diff_val, '', 0)}")
        if np.isfinite(diff_val): valid_count += 1
    log_list.append(f"    Values: [{', '.join(log_details)}]")
    if valid_count > 0:
        avg_ebitda_tax = np.nanmean(yearly_diffs)
        log_list.append(f"    Average ({valid_count} valid yrs): {format_currency(avg_ebitda_tax, '', 0)}")
        if np.isfinite(avg_ebitda_tax) and avg_ebitda_tax > 1e-9:
             log_list.append(f"  ✅ Found Positive Average. Using as projection basis.")
             return avg_ebitda_tax, f"Avg(EBITDA-TaxProv {valid_count}yr)"
        else: log_list.append(f"    Average is non-positive or invalid. Trying next method.")
    else: log_list.append(f"    No valid values for this method in the last {max_avg_years} years. Trying next method.")

    # Method 5 Avg: Avg(EBITDA Margin) * RevT0
    log_list.append(f"  Trying Method: Avg(EBITDAMargin)*RevT0...")
    fin_t0 = financials_df.iloc[:, 0]; cf_t0 = cashflow_df.iloc[:, 0]
    rev_t0 = safe_get(fin_t0, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    if np.isfinite(rev_t0) and rev_t0 > 1e-9:
        margins = []
        valid_count = 0
        log_details = []
        for i in range(max_avg_years):
            fin = financials_df.iloc[:, i]; cf = cashflow_df.iloc[:, i]
            margin_val = calc_method5(fin, None, cf, None) # Method 5 needs fin/cf
            margins.append(margin_val)
            log_details.append(f"T-{i}: {format_percent(margin_val)}")
            if np.isfinite(margin_val): valid_count += 1
        log_list.append(f"    Margins: [{', '.join(log_details)}]")
        if valid_count > 0:
             avg_margin = np.nanmean(margins)
             fcff_m5_basis = avg_margin * rev_t0
             log_list.append(f"    Average Margin ({valid_count} valid yrs): {format_percent(avg_margin)}, Rev T-0: {format_currency(rev_t0, '', 0)}")
             log_list.append(f"    Calculated FCFF Basis: {format_currency(fcff_m5_basis, '', 0)}")
             if np.isfinite(fcff_m5_basis) and fcff_m5_basis > 1e-9:
                 log_list.append(f"  ✅ Found Positive Basis. Using as projection basis.")
                 return fcff_m5_basis, f"Avg(EBITDAMargin {valid_count}yr)*RevT0"
             else: log_list.append(f"    Result is non-positive or invalid. No positive basis found.")
        else: log_list.append(f"    No valid margins found. No positive basis found.")
    else: log_list.append(f"    T-0 Revenue invalid ({format_currency(rev_t0)}). Cannot use AvgMargin*Rev method.")

    # --- Final Failure ---
    log_list.append(f"  ❌ No method yielded a positive average FCFF basis.")
    return np.nan, "No Positive Average FCFF Found"


# --- End Helper Functions ---


# %% [markdown]
# ## 3. User Input & Dynamic Fetching: Ticker, Rf, Rm, Benchmarks, Beta, Spread

# %%
input_sources = {}

# --- Get Ticker Symbol ---
user_ticker = input(f"Enter stock ticker (e.g., MSFT, AAPL, RELIANCE.NS) [Default: {DEFAULT_TICKER}]: ")
if not user_ticker: user_ticker = DEFAULT_TICKER
ticker_symbol = user_ticker.strip().upper()
display(Markdown(f"**Analyzing Ticker:** `{ticker_symbol}`"))

# ----> DETECT IF INDIAN STOCK for currency formatting <----
is_indian_stock = ticker_symbol.endswith(".NS")
if is_indian_stock:
    print("Info: Detected Indian stock (.NS). Using ₹/Lakh/Crore formatting.")

# --- Determine RF Ticker based on Stock Ticker ---
rf_proxy_ticker = "^TNX" # Default US 10Y
if is_indian_stock: rf_proxy_ticker = "^NS10"; print(f"Info: Using Indian Rf proxy: {rf_proxy_ticker}")
elif ticker_symbol.endswith(".L"): rf_proxy_ticker = "^FTLC"; print(f"Info: Using UK Rf proxy: {rf_proxy_ticker}")

# --- Get Risk-Free Rate (Rf) ---
print("\n--- CAPM Inputs ---")
dynamic_rf_rate = try_get_dynamic_rf(rf_proxy_ticker)
rf_prompt = f"Enter Risk-Free Rate (Rf) as percentage (e.g., '4.5' for 4.5%). Fetched {rf_proxy_ticker}?"
suggested_rf_val = dynamic_rf_rate * 100 if dynamic_rf_rate is not None else DEFAULT_RF_RATE_SUGGESTION
risk_free_rate_input_pct = get_validated_float_input(rf_prompt, default_value=f"{suggested_rf_val:.2f}", allow_blank=False, min_val=0.0, max_val=20.0)
risk_free_rate_input = risk_free_rate_input_pct / 100.0
input_sources['Rf'] = 'Dynamic Fetch (User Confirmed/Adjusted)' if dynamic_rf_rate is not None else f'User Input (Default: {DEFAULT_RF_RATE_SUGGESTION:.1f}%)'
display(Markdown(f"**Using Risk-Free Rate (Rf):** {format_percent(risk_free_rate_input)} (`Source: {input_sources['Rf']}`)"))

# --- Get Expected Market Return (Rm) ---
suggested_mrp = DEFAULT_MARKET_RISK_PREMIUM
suggested_rm = risk_free_rate_input + suggested_mrp
rm_prompt = f"Enter Expected Market Return (Rm) as percentage (e.g., '10'). Suggestion based on Rf + {format_percent(suggested_mrp)} MRP is {format_percent(suggested_rm)}"
while True:
    market_return_input_pct = get_validated_float_input(rm_prompt, default_value=f"{suggested_rm * 100:.1f}", allow_blank=False, min_val=-10.0, max_val=35.0)
    market_return_input = market_return_input_pct / 100.0
    if market_return_input >= risk_free_rate_input: break
    else: print(f"❌ Error: Rm ({format_percent(market_return_input)}) must be >= Rf ({format_percent(risk_free_rate_input)}). Re-enter.")
market_risk_premium_input = market_return_input - risk_free_rate_input
input_sources['Rm'] = 'User Input'; input_sources['MRP'] = 'Derived'
display(Markdown(f"**Using Expected Market Return (Rm):** {format_percent(market_return_input)} (`Source: {input_sources['Rm']}`)"))
display(Markdown(f"**↳ Implied Market Risk Premium (MRP):** {format_percent(market_risk_premium_input)} (`Source: {input_sources['MRP']}`)"))

# --- Beta (Placeholder) ---
beta_input = None; input_sources['Beta'] = 'Pending'
# --- Credit Spread (Placeholder) ---
user_credit_spread = None; input_sources['Credit Spread'] = 'Pending'

# --- Get Industry Benchmark Ratios (with fallback defaults) ---
print("\n--- Industry Ratio Benchmarks ---")
display(Markdown("Enter benchmarks or leave blank for generic fallbacks. *(Find industry averages online)*"))
user_benchmarks = {}; benchmark_sources = {}
print("Provide benchmarks (e.g., ROE: enter '15' for 15%; P/E: enter '20'):")
for ratio_key in REQUIRED_BENCHMARK_RATIOS:
    default_val = DEFAULT_BENCHMARK_RATIOS.get(ratio_key, None)
    prompt = f"  Enter benchmark for {ratio_key}"
    is_percent_type = ratio_key in ['ROE', 'ROA', 'ROCE']
    min_b, max_b = (-50, 100) if is_percent_type else (-100, 500)
    bench_val_input = get_validated_float_input(prompt, default_value=f"{default_val:.2f}" if default_val is not None else None, allow_blank=True, min_val=min_b, max_val=max_b)
    if bench_val_input is not None:
        user_benchmarks[ratio_key] = bench_val_input; benchmark_sources[ratio_key] = 'User Input'
        display_val = f"{bench_val_input:.1f}%" if is_percent_type else f"{bench_val_input:.2f}"
        display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: User`)"))
    elif default_val is not None:
         user_benchmarks[ratio_key] = default_val; benchmark_sources[ratio_key] = 'Fallback Default'
         display_val = f"{default_val:.1f}%" if is_percent_type else f"{default_val:.2f}"
         display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: Default`)"))
    else:
        benchmark_sources[ratio_key] = 'Not Available'
        display(Markdown(f"  - **{ratio_key} Benchmark:** Not Provided / No Default"))
input_sources['Benchmarks'] = 'User Input or Default'

# %% [markdown]
# ## 4. Data Acquisition (Fetching Extended History)

# %%
stock_data = {}
try:
    stock = yf.Ticker(ticker_symbol)
    info = stock.info
    if not info or not info.get('regularMarketPrice'):
        print(f"Warning: Limited info for {ticker_symbol}. Trying history...")
        hist_check = stock.history(period="5d");
        if hist_check.empty: raise ValueError(f"Ticker '{ticker_symbol}' invalid/no data.")
        else: print(f"Warning: Proceeding with limited data."); info = {} if not isinstance(info, dict) else info
    stock_data['info'] = info

    stock_data['financials'] = stock.financials
    stock_data['balance_sheet'] = stock.balance_sheet
    stock_data['cashflow'] = stock.cashflow
    if stock_data['financials'].empty or stock_data['balance_sheet'].empty or stock_data['cashflow'].empty:
       print(f"⚠️ Warning: Empty statements for {ticker_symbol}. DCF unreliable.")

    stock_data['history'] = stock.history(period="2y")
    stock_data['history_long'] = stock.history(period="4y", interval="1d")
    if not stock_data['history_long'].empty:
        print(f"Info: Fetched long history (4y daily). Columns available: {list(stock_data['history_long'].columns)}")
    else:
         print(f"⚠️ Warning: Could not fetch long history (4y daily) for {ticker_symbol}. Stock CAGR calc may fail.")
    if stock_data['history_long'].empty or len(stock_data['history_long']) < 2:
        print(f"⚠️ Warning: Fetched long history has < 2 data points. Stock CAGR calc will fail.")

    company_name = info.get("longName", info.get("shortName", ticker_symbol))
    currency_code = info.get("currency", DEFAULT_CURRENCY_SYMBOL)
    if not isinstance(currency_code, str) or len(currency_code)>5 : currency_code=DEFAULT_CURRENCY_SYMBOL
    currency_symbol = "₹" if is_indian_stock else info.get("symbol", currency_code)

    current_price = safe_get(info, ['currentPrice', 'regularMarketPrice', 'previousClose', 'lastPrice'], 'Info', default=np.nan)
    price_source = "yfinance Info"
    if not np.isfinite(current_price) and not stock_data['history'].empty:
        if 'Close' in stock_data['history'].columns and not stock_data['history']['Close'].empty:
            last_close_hist = stock_data['history']['Close'].iloc[-1]
            if pd.notna(last_close_hist): current_price = last_close_hist; price_source = "yfinance History (Close)"
        elif 'Adj Close' in stock_data['history'].columns and not stock_data['history']['Adj Close'].empty:
             last_adj_close_hist = stock_data['history']['Adj Close'].iloc[-1]
             if pd.notna(last_adj_close_hist): current_price = last_adj_close_hist; price_source = "yfinance History (Adj Close)"

    if not np.isfinite(current_price): print(f"⚠️ Warning: Could not determine current price."); current_price = np.nan

    shares_outstanding = safe_get(info, ['sharesOutstanding'], 'Info', default=0)

    beta_input = safe_get(info, ['beta'], 'Info', default=None)
    if beta_input is not None and np.isfinite(beta_input):
        input_sources['Beta'] = 'Dynamic Fetch'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))
    else:
         print(f"❌ ERROR: Beta not found dynamically or invalid.")
         beta_input = get_validated_float_input(f"Enter Beta for {ticker_symbol}:", allow_blank=False, min_val=-1.0, max_val=4.0)
         info['beta'] = beta_input; input_sources['Beta'] = 'User Input'
         display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))

    market_cap = safe_get(info, ['marketCap'], 'Info', default=0)
    mcap_source = "yfinance Info"
    if market_cap <= 0 and shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Calculated"; print(f"Info: Market Cap calculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}")
    elif market_cap <= 0: mcap_source = "Unavailable"; print(f"⚠️ Warning: Market Cap unavailable."); market_cap = np.nan

    sector = info.get("sector", "N/A"); industry = info.get("industry", "N/A")
    summary = info.get("longBusinessSummary", "No summary available.")

    display(Markdown(f"**Company:** {company_name}")); display(Markdown(f"**Sector:** {sector} | **Industry:** {industry}"))
    display(Markdown(f"**Currency:** {currency_code} ({currency_symbol})"));
    display(Markdown(f"**Current Market Price:** {format_currency(current_price, currency_symbol, is_indian=is_indian_stock)} (`Source: {price_source}`)"))
    display(Markdown(f"**Market Cap:** {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)} (`Source: {mcap_source}`)"))

except Exception as e: print(f"❌ CRITICAL ERROR fetching data: {e}"); sys.exit("Cannot proceed.")

if not isinstance(shares_outstanding, (int, float)) or shares_outstanding <= 0: sys.exit("Essential data missing: Shares Outstanding.")
if beta_input is None or not np.isfinite(beta_input): sys.exit("Essential data missing: Beta.")
if not np.isfinite(market_cap): print(f"⚠️ WARNING: Market Cap unavailable. WACC weights less accurate.")

# %% [markdown]
# ## 5. WACC Calculation

# %%
# WACC Function Definition (No changes needed here, already robust)
def calculate_wacc(stock_info_dict, financials_df, balance_sheet_df,
                   user_rf, user_mrp, user_beta, input_sources_dict,
                   currency_sym, is_indian): # Added currency params
    global user_credit_spread
    display(Markdown("### Calculating WACC..."))
    components = {}
    calculation_log = []
    calculation_log.append(f"Input Sources: Rf='{input_sources_dict.get('Rf')}', MRP='{input_sources_dict.get('MRP')}', Beta='{input_sources_dict.get('Beta')}'")
    beta_val = user_beta
    if beta_val <= 0: calculation_log.append(f"⚠️ Warning: Non-positive Beta ({beta_val:.2f}).")
    components['beta'] = beta_val; components['risk_free_rate'] = user_rf; components['market_risk_premium'] = user_mrp
    cost_of_equity = user_rf + beta_val * user_mrp
    components['cost_of_equity'] = cost_of_equity
    calculation_log.append(f"Inputs -> Rf: {format_percent(user_rf)}, Beta: {beta_val:.2f}, MRP: {format_percent(user_mrp)}")
    calculation_log.append(f"↳ Cost of Equity (Ke) [CAPM]: {format_percent(cost_of_equity)}")
    cost_of_debt_pre_tax = np.nan; tax_rate = np.nan; after_tax_cost_of_debt = np.nan
    total_debt = 0.0; debt_calculation_method = "N/A"; tax_rate_source = "N/A"
    if not balance_sheet_df.empty and not financials_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]; latest_fin = financials_df.iloc[:, 0]
        debt_keys = ['Total Debt', 'Long Term Debt', 'Long Term Debt And Capital Lease Obligation', 'Total Debt Non Current']
        long_term_debt = safe_get(latest_bs, debt_keys , 'Balance Sheet', default=0)
        current_debt_keys = ['Current Debt', 'Short Term Debt', 'Current Debt And Capital Lease Obligation', 'Short Long Term Debt', 'Total Debt Current']
        short_term_debt = safe_get(latest_bs, current_debt_keys , 'Balance Sheet', default=0)
        total_debt = long_term_debt + short_term_debt
        components['debt_value_used'] = total_debt
        interest_expense = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], 'Income Statement', default=0))
        implied_rate_calculated = False
        if total_debt > 1000 and interest_expense > 0:
             implied_rate = interest_expense / total_debt; implied_rate_calculated = True
             if MIN_COST_OF_DEBT <= implied_rate <= MAX_COST_OF_DEBT:
                  cost_of_debt_pre_tax = implied_rate; debt_calculation_method = "Implied";
                  calculation_log.append(f"Interest Exp: {format_currency(interest_expense, symbol=currency_sym, is_indian=is_indian)}, Debt: {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
                  calculation_log.append(f"↳ Implied Cost of Debt (Kd pre-tax): {format_percent(cost_of_debt_pre_tax)}")
             else: calculation_log.append(f"⚠️ Implied Kd ({format_percent(implied_rate)}) out of bounds. Falling back.")
        elif total_debt > 1000: calculation_log.append(f"ℹ️ Found Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}) but no Interest Exp. Falling back.")
        if not debt_calculation_method.startswith("Implied"):
             if total_debt > 1000:
                 debt_calculation_method = "Default (Rf+Spread)"
                 if user_credit_spread is None:
                     spread_prompt = f"Implied Kd failed/NA. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
                     user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
                     user_credit_spread = user_credit_spread_pct / 100.0
                     input_sources_dict['Credit Spread'] = 'User Input (Default Kd)'
                 else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
                 cost_of_debt_pre_tax = user_rf + user_credit_spread
                 calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(cost_of_debt_pre_tax)}")
                 calculation_log.append(f"   (Spread Source: {input_sources_dict['Credit Spread']})")
             else:
                 debt_calculation_method = "Zero Debt"; cost_of_debt_pre_tax = 0.0; input_sources_dict['Credit Spread'] = 'N/A (Zero Debt)';
                 calculation_log.append(f"ℹ️ Negligible Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}). Kd = 0%.")
        income_tax_expense = safe_get(latest_fin, ['Tax Provision', 'Income Tax Expense'], 'Income Statement', default=0)
        pretax_income = safe_get(latest_fin, ['Pretax Income', 'EBT', 'Earnings Before Tax'], 'Income Statement', default=0)
        if pretax_income <= 0 and income_tax_expense > 0:
             ebit_val = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=0); interest_exp_for_ebt = abs(safe_get(latest_fin, ['Interest Expense', 'Interest Expense Net'], default=0))
             pretax_income_from_ebit = ebit_val - interest_exp_for_ebt
             if pretax_income_from_ebit > 0: pretax_income = pretax_income_from_ebit
        if pretax_income > 0 and income_tax_expense >= 0:
            calculated_tax_rate = income_tax_expense / pretax_income; tax_rate = min(max(calculated_tax_rate, MIN_TAX_RATE), MAX_TAX_RATE); tax_rate_source = "Derived & Bounded"
            if abs(tax_rate - calculated_tax_rate) > 0.001: tax_rate_source = f"Derived ({format_percent(calculated_tax_rate)}), Bounded"
            calculation_log.append(f"Tax Exp: {format_currency(income_tax_expense, symbol=currency_sym, is_indian=is_indian)}, Pretax Inc: {format_currency(pretax_income, symbol=currency_sym, is_indian=is_indian)}")
            calculation_log.append(f"↳ Effective Tax Rate: {format_percent(tax_rate)} ({tax_rate_source})")
        else: tax_rate = DEFAULT_TAX_RATE; tax_rate_source = "Default"; calculation_log.append(f"⚠️ Cannot derive tax rate. Using default: {format_percent(tax_rate)}")
    else: # Financial statements empty
         calculation_log.append("⚠️ Financial statements empty. Using defaults."); debt_calculation_method = "Default (No Fin.)"; tax_rate_source = "Default (No Fin.)"
         if user_credit_spread is None:
              spread_prompt = f"No financials. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd (e.g., '3' for 3%)"
              user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=12.0)
              user_credit_spread = user_credit_spread_pct / 100.0; input_sources_dict['Credit Spread'] = 'User Input (No Fin.)'
         else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
         cost_of_debt_pre_tax = user_rf + user_credit_spread; tax_rate = DEFAULT_TAX_RATE; total_debt = 0; components['debt_value_used'] = 0
         calculation_log.append(f"Using Default Kd: {format_percent(cost_of_debt_pre_tax)} (Rf + Spread: {input_sources_dict['Credit Spread']})")
         calculation_log.append(f"Using Default Tax Rate: {format_percent(tax_rate)}")
    components['cost_of_debt'] = cost_of_debt_pre_tax if np.isfinite(cost_of_debt_pre_tax) else 0.0
    components['debt_calc_method'] = debt_calculation_method
    components['tax_rate'] = tax_rate if np.isfinite(tax_rate) else DEFAULT_TAX_RATE
    components['tax_rate_source'] = tax_rate_source
    if np.isfinite(cost_of_debt_pre_tax) and np.isfinite(tax_rate): after_tax_cost_of_debt = cost_of_debt_pre_tax * (1 - tax_rate)
    else: after_tax_cost_of_debt = components['cost_of_debt'] * (1 - DEFAULT_TAX_RATE)
    components['after_tax_cost_of_debt'] = after_tax_cost_of_debt
    if components['cost_of_debt'] > 0 : calculation_log.append(f"↳ Cost of Debt (After-tax): {format_percent(after_tax_cost_of_debt)}")
    market_cap_val = safe_get(stock_info_dict, ['marketCap'], 'Info', default=np.nan); equity_source = "Market Cap" if np.isfinite(market_cap_val) and market_cap_val > 0 else "N/A"
    equity_value_for_wacc = np.nan
    if np.isfinite(market_cap_val) and market_cap_val > 0: equity_value_for_wacc = market_cap_val
    else:
       calculation_log.append("⚠️ Market Cap unavailable/zero."); equity_source = "Book Equity Fallback"
       if not balance_sheet_df.empty:
           equity_keys = ['Total Stockholder Equity', 'Stockholders Equity', 'Total Equity Gross Minority Interest', 'Common Stock Equity']
           book_equity_val = safe_get(balance_sheet_df.iloc[:, 0], equity_keys, 'Balance Sheet', default=np.nan)
           if np.isfinite(book_equity_val) and book_equity_val > 1.0:
              equity_value_for_wacc = book_equity_val; calculation_log.append(f"Using Book Equity ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}) as proxy.")
           else: equity_source = "Placeholder (Book Invalid)"
       else: equity_source = "Placeholder (No BS)"
       if not np.isfinite(equity_value_for_wacc): equity_value_for_wacc = 1.0; calculation_log.append("⚠️ Market Cap & Book Equity unavailable. Using placeholder Equity Value (1.0). Weights inaccurate.")
    components['equity_value_used'] = equity_value_for_wacc; components['equity_source'] = equity_source
    equity_weight = np.nan; debt_weight = np.nan; weights_source = "N/A"; total_capital = equity_value_for_wacc + total_debt
    if np.isfinite(total_capital) and total_capital > 1.0:
        equity_weight = equity_value_for_wacc / total_capital; debt_weight = total_debt / total_capital; weights_source = "Calculated"
        total_weight = equity_weight + debt_weight
        if abs(total_weight - 1.0) > 1e-6: equity_weight /= total_weight; debt_weight /= total_weight
    else: calculation_log.append(f"❌ ERROR: Total Capital ({format_currency(total_capital, symbol=currency_sym, is_indian=is_indian)}) invalid. Using default weights (70/30)."); equity_weight = 0.7; debt_weight = 0.3; weights_source = "Default"
    components['equity_weight'] = equity_weight; components['debt_weight'] = debt_weight; components['weights_source'] = weights_source
    calculation_log.append(f"Equity Value Used ({components['equity_source']}): {format_currency(components['equity_value_used'], symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"Debt Value Used ({components['debt_calc_method']}): {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"↳ Weights E/V: {format_percent(equity_weight)}, D/V: {format_percent(debt_weight)} (Source: {weights_source})")
    wacc = np.nan
    if all(np.isfinite(v) for v in [equity_weight, cost_of_equity, debt_weight, after_tax_cost_of_debt]):
        wacc = (equity_weight * cost_of_equity) + (debt_weight * after_tax_cost_of_debt); original_wacc = wacc; wacc = min(max(wacc, MIN_WACC), MAX_WACC)
        bound_note = ""
        if abs(wacc - original_wacc) > 0.0001 : bound_note = f"(Bounded from {format_percent(original_wacc)})"; calculation_log.append(f"⚠️ WACC ({format_percent(original_wacc)}) bounded.")
        calculation_log.append(f"▶️ Calculated WACC: {format_percent(wacc)} {bound_note}")
    else: calculation_log.append(f"❌ ERROR: Cannot calculate final WACC due to invalid components."); wacc = np.nan
    components['final_wacc'] = wacc
    display(Markdown("```\n" + "\n".join(calculation_log) + "\n```"))
    if np.isfinite(wacc) and wacc > 0 and np.isfinite(equity_weight) and np.isfinite(debt_weight):
        equity_contrib = equity_weight * cost_of_equity; debt_contrib = debt_weight * after_tax_cost_of_debt
        if np.isfinite(equity_contrib) and np.isfinite(debt_contrib) and abs(equity_contrib + debt_contrib - wacc) < 0.01:
           labels = ['Cost of Equity Contribution', 'Cost of Debt Contribution']; values = [equity_contrib, debt_contrib]
           fig_wacc_donut = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4, marker_colors=px.colors.qualitative.Pastel, hovertemplate='<b>%{label}</b><br>Contribution: %{value:.2%}<extra></extra>', textinfo='percent', textfont_size=12, sort=False)])
           fig_wacc_donut.update_layout(title_text=f'WACC Composition (Total: {format_percent(wacc)})', title_x=0.5, showlegend=True, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), margin=dict(l=20, r=20, t=60, b=50), height=350)
           fig_wacc_donut.show()
        else: display(Markdown(f"*(WACC component contributions invalid/don't sum. Cannot generate chart.)*"))
    elif not np.isfinite(wacc): display(Markdown(f"*(WACC calculation failed. Cannot generate chart)*"))
    else: display(Markdown(f"*(WACC zero/negative or weights invalid. Cannot generate chart)*"))
    return wacc, components

# Execute WACC Calculation
discount_rate, wacc_components = calculate_wacc(
    stock_data['info'], stock_data['financials'], stock_data['balance_sheet'],
    risk_free_rate_input, market_risk_premium_input, beta_input, input_sources,
    currency_symbol, is_indian_stock
)

# %% [markdown]
# ## 6. FCFF Calculation & Projection (Initial Growth: Resilient GM + Positive Avg FCFF Basis V21)

# %%
# FCFF Function Definition (V21 - Positive Avg FCFF Basis Logic)
def calculate_and_project_fcff_v21(financials_df, balance_sheet_df, cashflow_df, history_long_df, # History needed for stock CAGR
                                   wacc_comps, disc_rate, forecast_yrs,
                                   final_user_rf, default_init_growth, # This is 6.5% default
                                   currency_sym, is_indian): # Added currency params
    display(Markdown("### Calculating & Projecting FCFF..."))
    projection_log = []
    projection_results = {
        'historical_fcffs_individual': {}, # Store individual yearly best estimates for CAGR calc
        'historical_revenues': {}, # Store recent historical revenues
        'current_fcff_basis': np.nan, # This will be the positive average basis
        'projected_fcffs': [], 'growth_rates': [],
        'pv_fcffs': [], 'cumulative_pv_fcff': np.nan, 'terminal_growth_rate': np.nan,
        'terminal_value': np.nan, 'pv_terminal_value': np.nan, 'error': None,
        'initial_growth_rate': np.nan, 'growth_source': 'N/A',
        'fcff_basis_method': 'N/A' # Store method used for the POSITIVE average basis
    }

    if financials_df.empty or balance_sheet_df.empty or cashflow_df.empty:
         projection_results['error'] = "Missing Financials/Balance Sheet/Cash Flow."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results
    if not np.isfinite(disc_rate):
         projection_results['error'] = "Invalid WACC provided."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results

    # --- Calculate Historical FCFF (Individual Years for CAGR) & Revenue ---
    num_years_cagr = 3
    max_hist_years = num_years_cagr + 1 # Need 4 data points for 3yr CAGR
    num_statement_years = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    historical_fcffs_list_individual = [] # Best estimate for each year T-0..T-3
    historical_revenues_list = [] # Newest first (index 0 = T-0)

    effective_tax_rate = DEFAULT_TAX_RATE
    if isinstance(wacc_comps, dict) and 'tax_rate' in wacc_comps and np.isfinite(wacc_comps['tax_rate']):
        effective_tax_rate = wacc_comps['tax_rate']

    projection_log.append(f"Calculating historical FCFF (Individual Best Estimate for CAGR) & Revenue (up to {max_hist_years} points)...")
    years_to_calc = min(num_statement_years, max_hist_years)
    if years_to_calc < max_hist_years:
         projection_log.append(f"  ⚠️ Warning: Only {years_to_calc} years of statement data available (needed {max_hist_years} for full CAGR calculations).")

    for i in range(years_to_calc):
        fin_col = financials_df.iloc[:, i]; bs_col = balance_sheet_df.iloc[:, i]; cf_col = cashflow_df.iloc[:, i]
        bs_prev_col = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None
        # Use the single-year fallback helper for the historical list needed for CAGR
        fcff_val, fcff_method_yr = _calculate_fcff_single_year_best_estimate(fin_col, bs_col, cf_col, bs_prev_col, effective_tax_rate)
        historical_fcffs_list_individual.append(fcff_val)
        projection_log.append(f"  FCFF T-{i} (for CAGR): {format_currency(fcff_val, '', 0)} (Method: {fcff_method_yr})")

        rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
        revenue_val = safe_get(fin_col, rev_keys, default=np.nan)
        historical_revenues_list.append(revenue_val)

    # Pad with NaNs if data is shorter than needed for CAGR calculations later
    while len(historical_fcffs_list_individual) < max_hist_years: historical_fcffs_list_individual.append(np.nan)
    while len(historical_revenues_list) < max_hist_years: historical_revenues_list.append(np.nan)

    projection_results['historical_fcffs_individual'] = {f"T-{i}": v for i, v in enumerate(historical_fcffs_list_individual)}
    projection_results['historical_revenues'] = {f"T-{i}": v for i, v in enumerate(historical_revenues_list)}

    # --- V21: Determine T-0 FCFF starting basis using positive average priority ---
    # Pass the existing log list to capture the attempts within the helper
    fcff_start_basis, fcff_basis_method = get_positive_average_fcff_basis_v21(financials_df, balance_sheet_df, cashflow_df, effective_tax_rate, projection_log)
    projection_results['current_fcff_basis'] = fcff_start_basis
    projection_results['fcff_basis_method'] = fcff_basis_method
    # Logging is handled inside the helper

    # --- V21: Halt if no POSITIVE average basis found ---
    if not np.isfinite(fcff_start_basis):
        error_msg = f"Could not determine a valid POSITIVE average FCFF basis for projection start (Final method tried: {fcff_basis_method}). Cannot proceed."
        projection_results['error'] = error_msg
        # No need to log again, already logged by helper
        display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
        return projection_results # Stop before growth calculation / projection

    # --- Initial Growth Rate Calculation (GM of valid 3yr CAGRs - V21) ---
    # Uses the INDIVIDUAL historical FCFFs calculated earlier
    projection_log.append("--- Determining Initial Growth Rate (Geometric Mean of valid 3yr CAGRs) ---")
    initial_growth_rate = np.nan # Start with NaN
    growth_source = "N/A"
    calculated_cagrs = {} # Store individual CAGRs {Name: value}
    valid_cagr_list = [] # Store only the valid CAGRs for GM calculation

    # 1. Calculate 3-Year Revenue CAGR (Conditional)
    rev_cagr = np.nan
    projection_log.append(f"Calculating Revenue {num_years_cagr}yr CAGR...")
    if len(historical_revenues_list) >= max_hist_years:
        rev_end = historical_revenues_list[0]; rev_start = historical_revenues_list[num_years_cagr]
        if pd.notna(rev_end) and math.isfinite(rev_end) and pd.notna(rev_start) and math.isfinite(rev_start) and rev_start > 1e-9:
            rev_cagr = calculate_cagr_modified(rev_start, rev_end, num_years_cagr)
            if pd.notna(rev_cagr) and math.isfinite(rev_cagr):
                 calculated_cagrs['Revenue'] = rev_cagr; valid_cagr_list.append(rev_cagr)
                 projection_log.append(f"  ✅ Revenue CAGR: {format_percent(rev_cagr)} (Start: {format_currency(rev_start, '', 0, is_indian)}, End: {format_currency(rev_end, '', 0, is_indian)})")
            else: projection_log.append(f"  ❌ Revenue CAGR Calculation Failed (Result Invalid). (S={format_currency(rev_start, '', 0, is_indian)}, E={format_currency(rev_end, '', 0, is_indian)})")
        else: projection_log.append(f"  ❌ Revenue CAGR Skipped (Invalid/Non-positive Start Data). (S={format_currency(rev_start, '', 0, is_indian)}, E={format_currency(rev_end, '', 0, is_indian)})")
    else: projection_log.append(f"  ❌ Revenue CAGR Skipped (Insufficient data: {len(historical_revenues_list)} points)")

    # 2. Calculate 3-Year FCFF CAGR (Conditional - uses INDIVIDUAL historical FCFFs)
    fcff_cagr = np.nan
    projection_log.append(f"Calculating FCFF {num_years_cagr}yr CAGR...")
    if len(historical_fcffs_list_individual) >= max_hist_years:
        fcff_end = historical_fcffs_list_individual[0]; fcff_start = historical_fcffs_list_individual[num_years_cagr]
        if pd.notna(fcff_end) and math.isfinite(fcff_end) and pd.notna(fcff_start) and math.isfinite(fcff_start) and fcff_start > 1e-9:
            fcff_cagr = calculate_cagr_modified(fcff_start, fcff_end, num_years_cagr)
            if pd.notna(fcff_cagr) and math.isfinite(fcff_cagr):
                 calculated_cagrs['FCFF'] = fcff_cagr; valid_cagr_list.append(fcff_cagr)
                 projection_log.append(f"  ✅ FCFF CAGR:    {format_percent(fcff_cagr)} (Start: {format_currency(fcff_start, '', 0, is_indian)}, End: {format_currency(fcff_end, '', 0, is_indian)})")
            else: projection_log.append(f"  ❌ FCFF CAGR Calculation Failed (Result Invalid). (S={format_currency(fcff_start, '', 0, is_indian)}, E={format_currency(fcff_end, '', 0, is_indian)})")
        else: projection_log.append(f"  ❌ FCFF CAGR Skipped (Invalid/Non-positive Start Data). (S={format_currency(fcff_start, '', 0, is_indian)}, E={format_currency(fcff_end, '', 0, is_indian)})")
    else: projection_log.append(f"  ❌ FCFF CAGR Skipped (Insufficient data: {len(historical_fcffs_list_individual)} points)")

    # 3. Calculate 3-Year Stock Price CAGR (Conditional - Robust Column V21)
    stock_cagr = np.nan; price_col = None
    projection_log.append(f"Calculating Stock Price {num_years_cagr}yr CAGR...")
    if history_long_df is not None and not history_long_df.empty and len(history_long_df) >= 2:
         if 'Adj Close' in history_long_df.columns: price_col = 'Adj Close'
         elif 'Close' in history_long_df.columns: price_col = 'Close'; projection_log.append("  ⚠️ Using 'Close' price for Stock CAGR ('Adj Close' not found).")
         else: projection_log.append(f"  ❌ Stock CAGR Skipped (Neither 'Adj Close' nor 'Close' columns found)")
         if price_col and isinstance(history_long_df.index, pd.DatetimeIndex):
             projection_log.append(f"  Stock History Info: Using '{price_col}' column. Found {len(history_long_df)} rows, Date Range: {history_long_df.index.min().date()} to {history_long_df.index.max().date()}")
             try:
                 end_date = history_long_df.index[-1]; end_price = history_long_df[price_col].iloc[-1]
                 start_date_target = end_date - pd.DateOffset(years=num_years_cagr)
                 temp_series = history_long_df.loc[start_date_target:]
                 if not temp_series.empty:
                      start_actual_date = temp_series.index[0]; start_price = temp_series[price_col].iloc[0]
                      actual_years = (end_date - start_actual_date).days / 365.25
                      projection_log.append(f"  Stock CAGR Data Points: Start Date={start_actual_date.date()}, Start Price={start_price:.2f}, End Date={end_date.date()}, End Price={end_price:.2f}, Actual Years={actual_years:.2f}")
                      is_end_valid = pd.notna(end_price) and math.isfinite(end_price)
                      is_start_valid = pd.notna(start_price) and math.isfinite(start_price) and start_price > 1e-9
                      is_period_valid = actual_years >= (num_years_cagr - 0.5) and actual_years > 0
                      if is_end_valid and is_start_valid and is_period_valid:
                          stock_cagr = calculate_cagr_modified(start_price, end_price, actual_years)
                          if pd.notna(stock_cagr) and math.isfinite(stock_cagr):
                              calculated_cagrs['Stock'] = stock_cagr; valid_cagr_list.append(stock_cagr)
                              projection_log.append(f"  ✅ Stock CAGR:   {format_percent(stock_cagr)} (Using '{price_col}')")
                          else: projection_log.append(f"  ❌ Stock CAGR Calculation Failed (Result Invalid).")
                      else:
                          reason = [];
                          if not is_end_valid: reason.append("Invalid End Price");
                          if not is_start_valid: reason.append("Invalid/Non-positive Start Price")
                          if not is_period_valid: reason.append(f"Period Invalid ({actual_years:.2f} yrs vs target {num_years_cagr})")
                          projection_log.append(f"  ❌ Stock CAGR Skipped ({', '.join(reason)}).")
                 else: projection_log.append(f"  ❌ Stock CAGR Skipped (Could not find data at or after target start date: {start_date_target.date()})")
             except Exception as e: projection_log.append(f"  ❌ Stock CAGR Skipped (Error during lookup/calculation: {e})")
         elif not price_col: pass
         else: projection_log.append(f"  ❌ Stock CAGR Skipped (History index is not DatetimeIndex)")
    else:
         reason = [];
         if history_long_df is None: reason.append("History DF is None")
         elif history_long_df.empty: reason.append("History DF is empty")
         elif len(history_long_df) < 2: reason.append("Less than 2 data points")
         projection_log.append(f"  ❌ Stock CAGR Skipped (Insufficient/invalid history data object: {', '.join(reason)})")

    # 4. Calculate Geometric Mean of the valid CAGRs
    if valid_cagr_list:
        gm_growth = geometric_mean_modified(valid_cagr_list)
        if pd.notna(gm_growth) and math.isfinite(gm_growth):
            initial_growth_rate = gm_growth; growth_source = f"GM of {len(valid_cagr_list)} valid CAGR(s)"
            projection_log.append(f"  ▶️ Calculated Geometric Mean = {format_percent(initial_growth_rate)}")
            included_cagrs = [f"{k}:{format_percent(v)}" for k, v in calculated_cagrs.items()]
            projection_log.append(f"     Based on: [{', '.join(included_cagrs)}]")
        else:
            projection_log.append(f"  ⚠️ Geometric Mean calculation failed or result invalid. Falling back to default.")
            initial_growth_rate = default_init_growth; growth_source = f"Default ({format_percent(default_init_growth)}) [GM Fail]"
            projection_log.append(f"     CAGRs calculated but GM failed: {valid_cagr_list}")
    else:
        projection_log.append(f"  ⚠️ No valid 3yr CAGRs could be calculated. Falling back to default.")
        initial_growth_rate = default_init_growth; growth_source = f"Default ({format_percent(default_init_growth)}) [No Valid CAGRs]"

    # 5. Apply Floor (1.1%) and Ceiling (15%)
    unbounded_rate = initial_growth_rate
    initial_growth_rate = max(initial_growth_rate, MIN_INITIAL_GROWTH_FLOOR); initial_growth_rate = min(initial_growth_rate, MAX_INITIAL_GROWTH_CEILING)
    if abs(initial_growth_rate - unbounded_rate) > 1e-6: growth_source += " (Bounded)"
    projection_log.append(f"  Applied Bounds: Final Initial Growth Rate = {format_percent(initial_growth_rate)} (Floor: {format_percent(MIN_INITIAL_GROWTH_FLOOR)}, Ceiling: {format_percent(MAX_INITIAL_GROWTH_CEILING)})")

    projection_results['initial_growth_rate'] = initial_growth_rate; projection_results['growth_source'] = growth_source

    # --- Terminal Growth Rate Calculation (Only if positive basis was found) ---
    g_cap_initial_valid = initial_growth_rate if pd.notna(initial_growth_rate) and math.isfinite(initial_growth_rate) else TERMINAL_GROWTH_RATE_CAP
    g_cap_config = TERMINAL_GROWTH_RATE_CAP; g_cap_wacc = disc_rate * 0.67; g_cap_rf = final_user_rf * 0.95
    projection_log.append(f"--- Determining Terminal Growth Rate ---")
    projection_log.append(f"Terminal Growth Caps Considered: Config={format_percent(g_cap_config)}, WACC*0.67={format_percent(g_cap_wacc)}, Initial(Final)={format_percent(g_cap_initial_valid)}, Rf*0.95={format_percent(g_cap_rf)}")
    terminal_growth_rate = min(g_cap_config, g_cap_wacc, g_cap_initial_valid, g_cap_rf); projection_log.append(f"↳ Minimum of Caps: {format_percent(terminal_growth_rate)}")
    original_g_before_floor = terminal_growth_rate
    terminal_growth_rate = max(MIN_GROWTH, terminal_growth_rate)
    if abs(terminal_growth_rate - original_g_before_floor) > 1e-6 : projection_log.append(f"↳ Adjusted by Floor ({format_percent(MIN_GROWTH)}): {format_percent(terminal_growth_rate)}")
    if terminal_growth_rate >= disc_rate: original_g = terminal_growth_rate; terminal_growth_rate = disc_rate - 0.005; projection_log.append(f"⚠️ Terminal Growth ({format_percent(original_g)}) >= WACC ({format_percent(disc_rate)}). Reduced to {format_percent(terminal_growth_rate)}.")
    projection_log.append(f"▶️ Using Final Terminal Growth Rate (g): {format_percent(terminal_growth_rate)}"); projection_results['terminal_growth_rate'] = terminal_growth_rate

    # --- Project FCFFs (Only if positive basis was found) ---
    projected_fcffs = []; growth_rates_applied = []; pv_fcffs = []; cumulative_pv_fcff = 0.0
    current_fcff_for_proj = fcff_start_basis # Use the POSITIVE AVERAGE basis

    if pd.isna(initial_growth_rate) or not math.isfinite(initial_growth_rate):
        projection_log.append(f"⚠️ FATAL: Invalid final initial growth rate ({initial_growth_rate}) despite checks. Using 0% for projection start.")
        effective_initial_growth = 0.0
    else: effective_initial_growth = initial_growth_rate

    effective_terminal_growth = terminal_growth_rate

    projection_log.append("--- FCFF Projections ---"); projection_log.append(f"{'Year':<5} {'Growth':<8} {'FCFF':<18} {'PV Factor':<10} {'PV FCFF':<18}"); projection_log.append("-" * 60)
    last_valid_fcff = current_fcff_for_proj

    for year in range(1, forecast_yrs + 1):
        if forecast_yrs > 0: decay_factor = max(0, (forecast_yrs - year)) / forecast_yrs
        else: decay_factor = 0
        growth_rate = effective_terminal_growth + (effective_initial_growth - effective_terminal_growth) * decay_factor
        growth_rate = max(MIN_GROWTH, growth_rate)

        growth_rates_applied.append(growth_rate)
        fcff_forecast = last_valid_fcff * (1 + growth_rate)
        if not np.isfinite(fcff_forecast):
             projection_log.append(f"⚠️ Year {year} FCFF forecast became non-finite (Start:{last_valid_fcff}, Growth:{growth_rate}). Setting to 0.")
             fcff_forecast = 0
        projected_fcffs.append(fcff_forecast)
        last_valid_fcff = fcff_forecast

        pv_factor = 1 / ((1 + disc_rate) ** year); pv_fcff = fcff_forecast * pv_factor
        if not np.isfinite(pv_fcff): pv_fcff = 0
        pv_fcffs.append(pv_fcff); cumulative_pv_fcff += pv_fcff
        projection_log.append(f"{year:<5} {format_percent(growth_rate):<8} {format_currency(fcff_forecast, '', 0, is_indian=is_indian):<18} {pv_factor:<10.4f} {format_currency(pv_fcff, '', 0, is_indian=is_indian):<18}")

    projection_results['projected_fcffs'] = projected_fcffs; projection_results['growth_rates'] = growth_rates_applied; projection_results['pv_fcffs'] = pv_fcffs
    projection_results['cumulative_pv_fcff'] = cumulative_pv_fcff if np.isfinite(cumulative_pv_fcff) else 0.0
    projection_log.append("-" * 60); projection_log.append(f"Sum of PV(Forecasted FCFFs): {format_currency(cumulative_pv_fcff, symbol=currency_sym, is_indian=is_indian)}")

    # --- Calculate Terminal Value (Only if projection occurred) ---
    terminal_value = np.nan; pv_terminal_value = np.nan
    final_forecast_fcff = projected_fcffs[-1] if projected_fcffs and np.isfinite(projected_fcffs[-1]) else 0

    if (disc_rate - terminal_growth_rate) <= 1e-6 or not np.isfinite(terminal_growth_rate):
         projection_log.append(f"❌ ERROR: WACC ({format_percent(disc_rate)}) <= Terminal Growth ({format_percent(terminal_growth_rate)}) or g invalid. TV calc failed."); projection_results['error'] = "WACC <= g or g invalid. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    elif not np.isfinite(final_forecast_fcff):
         projection_log.append(f"❌ ERROR: Final projected FCFF (Year {forecast_yrs}) is invalid. TV calc failed."); projection_results['error'] = "Final FCFF invalid. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    else:
         if final_forecast_fcff <= 0: # Final check still needed
             projection_log.append(f"ℹ️ FCFF Year {forecast_yrs} ({format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)}) <= 0. Setting TV=0.");
             terminal_value = 0.0; pv_terminal_value = 0.0
         else:
              fcff_n_plus_1 = final_forecast_fcff * (1 + terminal_growth_rate)
              terminal_value = fcff_n_plus_1 / (disc_rate - terminal_growth_rate)
              if not np.isfinite(terminal_value) or terminal_value < 0:
                  projection_log.append(f"⚠️ Calculated TV invalid/negative ({format_currency(terminal_value, currency_sym, is_indian=is_indian)}). Setting TV=0.")
                  terminal_value = 0.0; pv_terminal_value = 0.0
              else:
                   pv_factor_tv = 1 / ((1 + disc_rate) ** forecast_yrs); pv_terminal_value = terminal_value * pv_factor_tv
                   projection_log.append(f"FCFF Yr {forecast_yrs}: {format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)} -> FCFF Yr {forecast_yrs+1}: {format_currency(fcff_n_plus_1, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"Terminal Value (Yr {forecast_yrs}): {format_currency(terminal_value, currency_sym, is_indian=is_indian)}")
                   projection_log.append(f"PV Factor TV: {pv_factor_tv:.4f} -> PV(TV): {format_currency(pv_terminal_value, currency_sym, is_indian=is_indian)}")

    projection_results['terminal_value'] = terminal_value if np.isfinite(terminal_value) else 0.0
    projection_results['pv_terminal_value'] = pv_terminal_value if np.isfinite(pv_terminal_value) else 0.0

    display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
    return projection_results

# --- Execute FCFF calculation and projection ---
# Initialize fcff_results dictionary to ensure it's defined before the try block
fcff_results = {'error': 'Calculation not run yet.', 'initial_growth_rate': np.nan, 'growth_source':'N/A', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'N/A'}

if np.isfinite(discount_rate):
    try:
        # Use the V21 function name
        fcff_results = calculate_and_project_fcff_v21(
            stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'], stock_data['history_long'], # Pass history_long
            wacc_components, discount_rate, FORECAST_YEARS, risk_free_rate_input, DEFAULT_GROWTH_RATE,
            currency_symbol, is_indian_stock)
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during FCFF Projection: {e}**"))
         # Ensure fcff_results is still a dict with error info
         fcff_results = {'error': f'Runtime error in FCFF calc: {e}', 'initial_growth_rate': np.nan, 'growth_source':'Error', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'Error'}

else:
    display(Markdown("❌ **Skipping FCFF Projection due to invalid WACC.**"));
    # Ensure fcff_results is a dict even if skipped
    fcff_results = {'error': 'Invalid WACC.', 'initial_growth_rate': np.nan, 'growth_source':'N/A (WACC Invalid)', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'N/A'}


# %% [markdown]
# ## 7. Enterprise Value and Equity Value Calculation

# %%
# Valuation Function Definition (No changes needed here, already robust)
def calculate_final_valuation(fcff_data, balance_sheet_df, shares_outstanding_val, currency_sym, wacc_data, is_indian): # Added flag
    display(Markdown("### Final Valuation Calculation..."))
    valuation = {'enterprise_value': np.nan, 'equity_value': np.nan, 'intrinsic_value_per_share': np.nan, 'error': None}
    valuation_log = []

    # Ensure fcff_data is a dictionary before accessing keys
    if not isinstance(fcff_data, dict):
        error_msg = f"FCFF Data is not a dictionary (type: {type(fcff_data)}). Cannot proceed."
        valuation['error'] = error_msg; valuation_log.append(f"❌ ERROR: {error_msg}")
        display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    if fcff_data.get('error'):
        valuation['error'] = f"FCFF Error: {fcff_data.get('error', 'Unknown')}"; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    pv_forecast_sum = fcff_data.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_data.get('pv_terminal_value', np.nan)
    if not np.isfinite(pv_forecast_sum) or not np.isfinite(pv_tv):
         valuation['error'] = f"Invalid PV components (PV FCFF: {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}, PV TV: {format_currency(pv_tv, currency_sym, is_indian=is_indian)}). Cannot calculate EV."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    enterprise_value = pv_forecast_sum + pv_tv; valuation['enterprise_value'] = enterprise_value
    valuation_log.append(f"PV(Forecast FCFFs): {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"PV(Terminal Value):   {format_currency(pv_tv, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"↳ Enterprise Value (EV): {format_currency(enterprise_value, currency_sym, is_indian=is_indian)}")
    total_debt_latest = 0.0; cash_and_equivalents = 0.0; minority_interest = 0.0
    debt_source_info = '?';
    if isinstance(wacc_data, dict):
        debt_source_info = wacc_data.get('debt_calc_method', '?'); total_debt_latest = wacc_data.get('debt_value_used', 0.0)
    else: valuation_log.append(f"⚠️ WACC data invalid, cannot get debt used in WACC. Assuming 0.")
    if not balance_sheet_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        cash_keys = ["Cash And Cash Equivalents", "Cash Cash Equivalents And Short Term Investments", "Cash Financial", "Cash", "Cash And Short Term Investments"]; cash_and_equivalents = safe_get(latest_bs, cash_keys, "BS", default=0)
        valuation_log.append(f"  + Cash & Equiv (BS): {format_currency(cash_and_equivalents, currency_sym, is_indian=is_indian)}")
        minority_interest = safe_get(latest_bs, ["Minority Interest", "Noncontrolling Interest"], "BS", default=0)
        if minority_interest != 0.0: valuation_log.append(f"  - Minority Interest: {format_currency(minority_interest, currency_sym, is_indian=is_indian)}")
    else:
        valuation_log.append("⚠️ Balance Sheet empty.")
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        valuation_log.append(f"  + Cash & Equiv: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")
    if not all(np.isfinite(v) for v in [enterprise_value, total_debt_latest, cash_and_equivalents, minority_interest]):
         valuation['error'] = "Invalid component for EV->Equity bridge."; valuation_log.append(f"❌ ERROR: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    equity_value = enterprise_value - total_debt_latest + cash_and_equivalents - minority_interest; valuation['equity_value'] = equity_value
    valuation_log.append(f"↳ Equity Value: {format_currency(equity_value, currency_sym, is_indian=is_indian)}")
    intrinsic_value_per_share = np.nan
    if not np.isfinite(equity_value): valuation['error'] = f"Equity value invalid ({equity_value})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    elif shares_outstanding_val <= 0: valuation['error'] = f"Shares outstanding invalid ({shares_outstanding_val})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    else:
        intrinsic_value_per_share = equity_value / shares_outstanding_val;
        valuation_log.append(f"  ÷ Shares Out: {shares_outstanding_val:,.0f}")
        valuation_log.append(f"▶️ Intrinsic Value/Share: {format_currency(intrinsic_value_per_share, currency_sym, is_indian=is_indian)}")
    valuation['intrinsic_value_per_share'] = intrinsic_value_per_share if np.isfinite(intrinsic_value_per_share) else np.nan
    display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
    return valuation

# --- Execute Final Valuation ---
final_valuation = {'error': 'Calculation not run yet.', 'intrinsic_value_per_share': np.nan}
if isinstance(fcff_results, dict) and isinstance(wacc_components, dict) and not fcff_results.get('error'):
    try:
        final_valuation = calculate_final_valuation(fcff_results, stock_data['balance_sheet'], shares_outstanding, currency_symbol, wacc_components, is_indian_stock)
    except Exception as e:
        display(Markdown(f"❌ **CRITICAL ERROR during Final Valuation: {e}**"))
        final_valuation = {'error': f'Runtime error in final valuation: {e}', 'intrinsic_value_per_share': np.nan}
elif isinstance(fcff_results, dict):
     display(Markdown(f"❌ **Skipping Final Valuation due to FCFF errors.** Error: {fcff_results.get('error')}"));
     final_valuation = {'error': fcff_results.get('error', 'Upstream FCFF error.'), 'intrinsic_value_per_share': np.nan}
else:
     display(Markdown("❌ **Skipping Final Valuation due to critical upstream errors (invalid data structures).**"));
     final_valuation = {'error': 'FCFF or WACC results dictionary not created/valid.', 'intrinsic_value_per_share': np.nan}


# %% [markdown]
# ## 8. Financial Ratio Analysis

# %%
# Ratio Function Definition (No changes needed here, already robust)
def analyze_ratios(stock_info_dict, financials_df, balance_sheet_df, cashflow_df,
                   enterprise_val, current_price_val, shares_outstanding_val,
                   currency_sym, user_benchmark_dict, # Only user inputs
                   benchmark_source_dict, # Dict tracking source per ratio
                   wacc_data, is_indian): # Pass WACC data for consistent debt and flag

    """ Calculates ratios, compares to user/fallback benchmarks, uses WACC-consistent debt. """
    display(Markdown("### Financial Ratio Analysis..."))
    ratios = {} # Calculated ratios {RatioName: value}
    ratio_log = []
    ratio_data_for_table = [] # For HTML table display
    ratio_plotly_data = {'Ratio': [], 'Company': [], 'Benchmark': [], 'Benchmark Source': [], 'Status': []} # For Plotly chart
    company_nm = stock_info_dict.get("shortName", ticker_symbol)

    # --- Pre-checks ---
    valid_price = current_price_val is not None and np.isfinite(current_price_val) and current_price_val > 0
    valid_shares = shares_outstanding_val is not None and np.isfinite(shares_outstanding_val) and shares_outstanding_val > 0
    valid_enterprise_val = enterprise_val is not None and np.isfinite(enterprise_val)

    # Check for essential statements
    if financials_df.empty or balance_sheet_df.empty:
         ratio_log.append("❌ ERROR: Income Statement or Balance Sheet empty. Cannot calculate most ratios.")
         display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
         return {}, [] # Return empty results

    latest_fin = financials_df.iloc[:, 0]; latest_bs = balance_sheet_df.iloc[:, 0]; latest_cf = cashflow_df.iloc[:, 0] if not cashflow_df.empty else pd.Series(dtype=float)

    # --- Calculate Ratio Components (using safe_get with NaN default where appropriate) ---
    net_income = safe_get(latest_fin, ["Net Income Continuous Operations", "Net Income", "Net Income Common Stockholders", "Net Income Applicable To Common Shares"], default=np.nan)
    revenue = safe_get(latest_fin, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
    depreciation = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], default=0) # Default 0 ok here
    ebitda = (ebit + depreciation) if np.isfinite(ebit) and np.isfinite(depreciation) else np.nan

    total_equity = safe_get(latest_bs, ["Total Stockholder Equity", "Stockholders Equity", "Total Equity Gross Minority Interest", "Common Stock Equity"], default=np.nan)
    total_assets = safe_get(latest_bs, ["Total Assets"], default=np.nan)
    current_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], default=np.nan)
    current_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], default=np.nan)
    inventory = safe_get(latest_bs, ["Inventory"], default=0) # Default 0 ok here

    # --- Use Debt value consistent with WACC calculation ---
    total_debt_ratio = 0.0; debt_source_info = '?'
    if isinstance(wacc_data, dict):
        total_debt_ratio = wacc_data.get('debt_value_used', 0.0)
        debt_source_info = wacc_data.get('debt_calc_method', '?');
    else:
        ratio_log.append(f"⚠️ WACC data invalid, cannot get debt used in WACC. Assuming 0 for ratio calcs.")
    ratio_log.append(f"Using Debt = {format_currency(total_debt_ratio, currency_sym, is_indian=is_indian)} (Source: WACC - {debt_source_info})")

    # Per Share Metrics
    eps = (net_income / shares_outstanding_val) if valid_shares and np.isfinite(net_income) and net_income != 0 else np.nan
    bvps = (total_equity / shares_outstanding_val) if valid_shares and np.isfinite(total_equity) and total_equity != 0 else np.nan

    # Capital Employed (ROCE denominator)
    capital_employed = np.nan; ce_source = "N/A"
    if np.isfinite(total_assets) and np.isfinite(current_liab):
        capital_employed_def1 = total_assets - current_liab
        if np.isfinite(capital_employed_def1) and capital_employed_def1 > 1.0:
            capital_employed = capital_employed_def1; ce_source = "A-CL"
    if not np.isfinite(capital_employed):
        if np.isfinite(total_equity) and np.isfinite(total_debt_ratio):
             capital_employed_def2 = total_equity + total_debt_ratio
             if np.isfinite(capital_employed_def2) and capital_employed_def2 > 1.0:
                 capital_employed = capital_employed_def2; ce_source = "E+D"
    if not np.isfinite(capital_employed):
         ce_source = "Invalid/Unavailable"; ratio_log.append("⚠️ Capital Employed could not be reliably calculated. ROCE affected.")

    # --- Calculate Ratios (Result is NaN if components invalid/zero denominator) ---
    ratios['P/E'] = (current_price_val / eps) if valid_price and np.isfinite(eps) and eps != 0 else np.nan
    ratios['P/B'] = (current_price_val / bvps) if valid_price and np.isfinite(bvps) and bvps != 0 else np.nan
    ratios['EV/EBITDA'] = (enterprise_val / ebitda) if valid_enterprise_val and np.isfinite(ebitda) and ebitda != 0 else np.nan
    ratios['ROE'] = (net_income / total_equity) if np.isfinite(net_income) and np.isfinite(total_equity) and total_equity != 0 else np.nan
    ratios['ROA'] = (net_income / total_assets) if np.isfinite(net_income) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['ROCE'] = (ebit / capital_employed) if np.isfinite(ebit) and np.isfinite(capital_employed) and capital_employed != 0 else np.nan
    ratios['Asset Turnover'] = (revenue / total_assets) if np.isfinite(revenue) and np.isfinite(total_assets) and total_assets != 0 else np.nan
    ratios['Current Ratio'] = (current_assets / current_liab) if np.isfinite(current_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan
    quick_assets = (current_assets - inventory) if np.isfinite(current_assets) and np.isfinite(inventory) else np.nan
    ratios['Quick Ratio'] = (quick_assets / current_liab) if np.isfinite(quick_assets) and np.isfinite(current_liab) and current_liab != 0 else np.nan

    # --- Compare with User/Fallback Benchmarks & Prepare Data Tables ---
    display(Markdown("*(Comparing ratios to user/fallback benchmarks...)*"))
    for key in REQUIRED_BENCHMARK_RATIOS:
        company_value = ratios.get(key); is_percent_type = key in ['ROE', 'ROA', 'ROCE']
        value_fmt = format_percent(company_value, 1) if is_percent_type else format_currency(company_value, '', 2, is_indian=False)
        if value_fmt == "N/A" and company_value is not None and not math.isfinite(company_value): value_fmt = "Inf"

        user_benchmark = user_benchmark_dict.get(key, None); fallback_benchmark = DEFAULT_BENCHMARK_RATIOS.get(key, None)
        benchmark_source = benchmark_source_dict.get(key, "?"); final_benchmark = None
        if benchmark_source == 'User Input' and user_benchmark is not None: final_benchmark = user_benchmark
        elif benchmark_source == 'Fallback Default' and fallback_benchmark is not None: final_benchmark = fallback_benchmark
        else: benchmark_source = "Not Available"

        bench_fmt = "N/A"
        if final_benchmark is not None: bench_fmt = f"{final_benchmark:.1f}%" if is_percent_type else f"{final_benchmark:.2f}"; bench_fmt += f" <i style='font-size:0.8em; color:grey'>({benchmark_source.replace('Fallback ','')})</i>"
        else: bench_fmt = "<i style='color:grey'>Not Provided</i>"

        status = "<i style='color:grey'>N/A</i>"; status_color = "grey"
        if np.isfinite(company_value) and final_benchmark is not None and np.isfinite(final_benchmark):
            numeric_value = company_value; numeric_benchmark = final_benchmark
            good_color = '#2ecc71'; bad_color = '#e74c3c'; neutral_color = '#f39c12'; grey_color='#7f8c8d'
            if abs(numeric_benchmark) > 1e-9: ratio_diff = (numeric_value - numeric_benchmark) / abs(numeric_benchmark)
            else: ratio_diff = float('inf') if numeric_value > 0 else float('-inf') if numeric_value < 0 else 0.0
            if key in ['ROE', 'ROA', 'ROCE', 'Asset Turnover', 'Current Ratio', 'Quick Ratio']: # Higher is better
                 if ratio_diff > 0.20: status_color = good_color; status = f"⬆️ Strong"
                 elif ratio_diff < -0.20: status_color = bad_color; status = f"⬇️ Weak"
                 else: status_color = grey_color; status = f"~ Average"
            elif key in ['P/E', 'P/B', 'EV/EBITDA']: # Lower might be better (valuation) or neutral
                 if ratio_diff < -0.20: status_color = good_color; status = f"✅ Low"
                 elif ratio_diff > 0.20: status_color = neutral_color; status = f"⚠️ High"
                 else: status_color = grey_color; status = f"~ Average"
            status = f"<span style='color:{status_color}'>{status}</span>"
            plot_value = numeric_value * 100 if is_percent_type else numeric_value; plot_benchmark = numeric_benchmark
            ratio_plotly_data['Ratio'].append(key); ratio_plotly_data['Company'].append(plot_value); ratio_plotly_data['Benchmark'].append(plot_benchmark);
            ratio_plotly_data['Benchmark Source'].append(benchmark_source); ratio_plotly_data['Status'].append(status)
        elif not np.isfinite(company_value): status = "<i style='color:grey'>Company N/A</i>"
        elif final_benchmark is None: status = "<i style='color:grey'>No Benchmark</i>"
        ratio_data_for_table.append([f"<b>{key}</b>", value_fmt, bench_fmt, status])

    # --- Display Logs & Table ---
    if ratio_log: display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
    if ratio_data_for_table:
        header = ["Ratio", f"{company_nm} Value", "Benchmark (Source)", "Status"]; html_table = "<p>(*Status vs User/Default benchmarks. Threshold: +/- 20%.)</p>"; html_table += "<table border='1' style='border-collapse:collapse; width:98%; margin-left:auto; margin-right:auto; font-size: 0.9em;'>"; html_table += "<thead style='background-color:#e9ecef; color: #495057;'><tr>" + "".join(f"<th style='padding: 8px; text-align:center;'>{h}</th>" for h in header) + "</tr></thead>"; html_table += "<tbody>"
        for i, row in enumerate(ratio_data_for_table): bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff"; html_table += f"<tr style='background-color:{bg_color};'>"; html_table += f"<td style='padding: 6px; text-align:left;'>{row[0]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[1]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[2]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[3]}</td>"; html_table += "</tr>"
        html_table += "</tbody></table><br>"; display(HTML(html_table))
    else: display(Markdown("*No ratio data available.*"))

    # --- Display Ratio Comparison Chart (Plotly Bullet) ---
    if ratio_plotly_data['Ratio']:
       num_ratios = len(ratio_plotly_data['Ratio']); fig_ratios = go.Figure()
       def get_range_colors(k): return ["#fadbd8", "#fdebd0", "#d5f5e3"] if k in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio'] else ["#d5f5e3","#fdebd0","#fadbd8"] if k in ['P/E','P/B','EV/EBITDA'] else ['lightgrey','darkgrey','grey']
       for i in range(num_ratios):
           ratio_name = ratio_plotly_data['Ratio'][i]; company_val = ratio_plotly_data['Company'][i]; benchmark_val = ratio_plotly_data['Benchmark'][i]; bench_src_plot = ratio_plotly_data['Benchmark Source'][i]
           if not np.isfinite(company_val) or not np.isfinite(benchmark_val): print(f"Skipping plot for {ratio_name}: Invalid Value ({company_val}) or Bench ({benchmark_val})."); continue
           range_colors = get_range_colors(ratio_name); is_percent = ratio_name in ['ROE', 'ROA', 'ROCE']
           poor_f=0.8; good_f=1.2; range1_limit = benchmark_val * (poor_f if benchmark_val >=0 else good_f); range2_limit = benchmark_val * (good_f if benchmark_val >=0 else poor_f)
           abs_max = max(abs(company_val), abs(benchmark_val)); plot_max_val = abs_max * 1.7 + 1; plot_min_val = 0
           if company_val < 0 or benchmark_val < 0: plot_min_val = min(company_val, benchmark_val) * 1.7 - 1
           if ratio_name in ['P/E', 'EV/EBITDA']: plot_max_val = min(plot_max_val, 100); plot_min_val = max(plot_min_val, -20)
           elif ratio_name in ['P/B']: plot_max_val = min(plot_max_val, 20); plot_min_val = max(plot_min_val, -5)
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: plot_max_val = min(plot_max_val, 5); plot_min_val = max(plot_min_val, 0)
           elif is_percent: plot_max_val = min(plot_max_val, 75); plot_min_val= max(plot_min_val,-50)
           else: plot_max_val = min(plot_max_val, 1000)
           if plot_min_val >= plot_max_val: plot_max_val = plot_min_val + 1
           steps = [];
           if abs(benchmark_val) < 1e-9: steps = [{'range': [plot_min_val, plot_max_val], 'color':'lightgrey'}]
           else: step_ranges = sorted([plot_min_val, range1_limit, range2_limit, plot_max_val]); step_colors = range_colors if benchmark_val >= 0 else range_colors[::-1]; steps = [{'range': [step_ranges[j], step_ranges[j+1]], 'color': step_colors[j]} for j in range(3)]
           bench_fmt_title = f"{benchmark_val:.1f}%" if is_percent else f"{benchmark_val:.2f}"; title_text = f"<b>{ratio_name}</b><br><i style='font-size:0.8em'>(Bench: {bench_fmt_title} [{bench_src_plot.replace('Fallback ','')}])</i>"
           lower_is_better = ratio_name in ['P/E', 'P/B', 'EV/EBITDA']; delta_decreasing_color = good_color if lower_is_better else bad_color; delta_increasing_color = bad_color if lower_is_better else good_color
           fig_ratios.add_trace(go.Indicator(
                mode = "number+gauge+delta", value = company_val, number={'suffix': "%" if is_percent else "", 'valueformat': ".1f"},
                delta = {'reference': benchmark_val, 'relative': True, 'valueformat':".0%", 'decreasing': {'color': delta_decreasing_color}, 'increasing': {'color': delta_increasing_color}},
                domain = {'row': i, 'column': 0}, title = {'text' : title_text, 'font':{'size':11}},
                gauge = { 'shape': "bullet", 'axis': {'range': [plot_min_val, plot_max_val] if plot_min_val < plot_max_val else [plot_min_val, plot_min_val+1], 'tickformat':'.1f'},
                    'threshold': { 'line': {'color': "black", 'width': 2}, 'thickness': 0.8, 'value': benchmark_val },
                    'steps': [ s for s in steps if s and isinstance(s.get('range'), list) and len(s['range'])==2 and s['range'][0] < s['range'][1] ],
                    'bar': {'color': "#2c3e50", 'thickness': 0.5} }
            ))
       fig_ratios.update_layout(height = max(300, num_ratios * 85), grid = {'rows': num_ratios, 'columns': 1, 'pattern': "independent"}, title_text='Ratio Performance vs. Benchmarks', title_x=0.5, margin=dict(l=160, r=40, t=60, b=30))
       fig_ratios.show()
    else: display(Markdown("*No valid ratios/benchmarks for comparison chart.*"))

    return ratios, ratio_data_for_table

# --- Execute Ratio Analysis ---
ratios_calculated = {}; ratio_table = []
if isinstance(final_valuation, dict) and isinstance(wacc_components, dict) and not final_valuation.get('error'):
    try:
        ev_for_ratios = final_valuation.get('enterprise_value', np.nan)
        ratios_calculated, ratio_table = analyze_ratios(
            stock_data['info'], stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'],
            ev_for_ratios, current_price, shares_outstanding, currency_symbol,
            user_benchmarks, benchmark_sources, wacc_components, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Ratio Analysis: {e}**"))
         ratios_calculated = {}; ratio_table = []
else:
    display(Markdown("❌ **Skipping Ratio Analysis due to prior errors or invalid data structures.**"));
    ratios_calculated = {}; ratio_table = []


# %% [markdown]
# ## 9. Summary & Visualization Dashboard

# %%
# Dashboard Function Definition (V21: Updated negative value context)
def display_summary_dashboard(company_nm, ticker_sym, current_pr, intrinsic_val_per_sh,
                              wacc_val, term_growth, init_growth, growth_src,
                              final_val_dict, hist_df, fcff_res, currency_sym, info_dict, input_src_dict,
                              is_indian): # Added flag

    display(Markdown(f"## Valuation Summary Dashboard: {company_nm} ({ticker_sym})"))
    source_summary = " | ".join(f"{k}:`{v.split('(')[0].strip()}`" for k, v in input_src_dict.items())
    growth_src_display = growth_src if growth_src else 'N/A'
    # Include the FCFF Basis Method in the sources if available
    fcff_basis_method_disp = fcff_res.get('fcff_basis_method', 'N/A') if isinstance(fcff_res, dict) else 'N/A'
    display(Markdown(f"**Input Sources:** {source_summary} | Growth:`{growth_src_display}` | FCFF Basis:`{fcff_basis_method_disp}`"))


    valid_intrinsic = intrinsic_val_per_sh is not None and np.isfinite(intrinsic_val_per_sh)
    valid_current_pr = current_pr is not None and np.isfinite(current_pr) and current_pr > 0
    valid_wacc = wacc_val is not None and np.isfinite(wacc_val)
    valid_term_g = term_growth is not None and np.isfinite(term_growth)
    valid_init_g = init_growth is not None and np.isfinite(init_growth)

    # --- Indicator Subplots ---
    fig_indicators = make_subplots(rows=1, cols=3, specs=[[{'type': 'indicator'}, {'type': 'indicator'}, {'type': 'indicator'}]],
                                 subplot_titles=("Price vs. Intrinsic", "WACC", "Growth Rates"), horizontal_spacing=0.1, column_widths=[0.4, 0.3, 0.3])

    # Indicator 1: Price vs Intrinsic
    intrinsic_val_num = intrinsic_val_per_sh if valid_intrinsic else np.nan
    intrinsic_val_fmt = format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian) if valid_intrinsic else "N/A"
    current_pr_num = current_pr if valid_current_pr else 0.0; upside_fmt = ""; delta_config = None
    comparison_text = f"Intrinsic: {intrinsic_val_fmt}" # Start with base text
    if valid_intrinsic and valid_current_pr :
        if intrinsic_val_num > 0:
             upside_pct = (intrinsic_val_num / current_pr_num) - 1
             upside_fmt = f" ({upside_pct:+.1%})"
             comparison_text += upside_fmt # Add upside % if intrinsic > 0
             delta_config = {'reference': intrinsic_val_num, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym, 'decreasing': {'color': '#2ecc71', 'symbol': '▲ Upside'}, 'increasing': {'color': '#e74c3c', 'symbol': '▼ Downside'}}
        elif intrinsic_val_num <= 0:
             comparison_text += " (Negative/Zero)" # Indicate non-positive intrinsic
             delta_config = {'reference': 0, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym, 'decreasing': {'color': 'orange', 'symbol': ''}, 'increasing': {'color': 'grey', 'symbol': ''}}
    elif valid_intrinsic: comparison_text += " (Current N/A)"
    else: comparison_text = f"Intrinsic: N/A"
    fig_indicators.add_trace(go.Indicator(mode = "number+delta" if delta_config else "number", value = current_pr_num if valid_current_pr else 0, number={'prefix': currency_sym + " " if is_indian else currency_sym, 'valueformat': '.2f', 'font':{'size':24}}, title = {"text": f"<span style='font-size:0.8em'>{comparison_text}</span>"}, delta = delta_config, domain = {'row': 0, 'column': 0}), row=1, col=1)

    # Indicator 2: WACC Gauge
    wacc_val_num = wacc_val * 100 if valid_wacc else None
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = wacc_val_num,
        number = {'suffix': "%", 'valueformat':'.1f' if valid_wacc else '', 'font':{'size':20}},
        gauge = {'axis': {'range': [MIN_WACC*100, MAX_WACC*100]}, 'bar': {'color': "darkblue" if valid_wacc else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc", 'steps' : [ {'range': [MIN_WACC*100, 9], 'color': "#d5f5e3"}, {'range': [9, 14], 'color': "#fdebd0"}, {'range': [14, MAX_WACC*100], 'color': "#fadbd8"}]} if valid_wacc else None,
        title = {"text": "<span style='font-size:0.8em'>WACC</span>"}, domain = {'row': 0, 'column': 1}), row=1, col=2)

    # Indicator 3: Growth Gauge
    term_growth_num = term_growth * 100 if valid_term_g else None; init_growth_fmt = format_percent(init_growth, 1) if valid_init_g else "N/A"
    growth_title_text = f"<span style='font-size:0.8em'>Terminal Growth (g)<br>(Initial: {init_growth_fmt})</span>"
    gauge_max_range = max(TERMINAL_GROWTH_RATE_CAP * 100 + 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) + 1)
    gauge_min_range = MIN_GROWTH * 100 - 1
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = term_growth_num,
         number = {'suffix': "%", 'valueformat':'.1f' if valid_term_g else '', 'font':{'size':20}},
         gauge = {'axis': {'range': [gauge_min_range, gauge_max_range], 'tickwidth': 1}, 'bar': {'color': "darkgrey" if valid_term_g else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [{'range': [gauge_min_range, 2.0], 'color': "lightcoral"}, {'range': [2.0, 4.5], 'color': "lightyellow"}, {'range': [4.5, gauge_max_range], 'color': "lightgreen"}],
                 'threshold' : {'line': {'color': "black", 'width': 3}, 'thickness': 0.75, 'value': term_growth_num } if valid_term_g else None } if valid_term_g else None,
         title = {"text": growth_title_text}, domain = {'row': 0, 'column': 2}), row=1, col=3)

    fig_indicators.update_layout(height=220, margin=dict(l=30, r=30, t=50, b=20)); fig_indicators.show()

    # --- Historical Price Chart ---
    if not hist_df.empty:
        price_col_hist = 'Close' if 'Close' in hist_df.columns else None
        if price_col_hist:
             fig_hist = px.line(hist_df, x=hist_df.index, y=price_col_hist, title=f'{company_nm} Price History vs. DCF'); fig_hist.update_layout(xaxis_title='Date', yaxis_title=f'Price ({currency_sym})', height=400, legend_title_text='Legend'); fig_hist.update_xaxes(rangeslider_visible=False)
             if valid_intrinsic: fig_hist.add_hline(y=intrinsic_val_num, line_dash="dash", line_color="#e74c3c", opacity=0.8, annotation_text=f"DCF Intrinsic: {format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian)}", annotation_position="bottom right", annotation_font_color="#e74c3c")
             if valid_current_pr: fig_hist.add_hline(y=current_pr_num, line_dash="dot", line_color="#3498db", opacity=0.8, annotation_text=f"Current: {format_currency(current_pr_num, currency_sym, is_indian=is_indian)}", annotation_position="top left", annotation_font_color="#3498db")
             fig_hist.show()
        else: display(Markdown("*Historical price data ('Close') unavailable for chart.*"))
    else: display(Markdown("*Historical price data unavailable.*"))

    # --- DCF EV Breakdown Pie Chart ---
    if isinstance(fcff_res, dict):
        pv_forecast_sum = fcff_res.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_res.get('pv_terminal_value', np.nan)
        valid_pvs = np.isfinite([pv_forecast_sum, pv_tv]); total_ev = (pv_forecast_sum + pv_tv) if all(valid_pvs) else np.nan
        if np.isfinite(total_ev) and not fcff_res.get('error'):
            labels_pie = []; values_pie = []; pull_pie = []; colors = ['#4e79a7','#f28e2b']
            if abs(pv_forecast_sum) > 1e-6: labels_pie.append(f'PV({FORECAST_YEARS}yr FCFFs)'); values_pie.append(pv_forecast_sum); pull_pie.append(0)
            if abs(pv_tv) > 1e-6: labels_pie.append('PV(TV)'); values_pie.append(pv_tv); pull_pie.append(0.05 if len(labels_pie) > 1 else 0)
            if labels_pie:
                hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                if is_indian: hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                fig_dcf_pie = go.Figure(data=[go.Pie(labels=labels_pie, values=values_pie, hole=.4, pull=pull_pie, marker_colors=colors[:len(labels_pie)], hovertemplate=hover_format, textinfo='percent+label', insidetextorientation='radial', sort=False)])
                fig_dcf_pie.update_layout( title_text=f"DCF EV Breakdown (Total EV: {format_currency(total_ev, currency_sym, is_indian=is_indian)})", title_x=0.5, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), height=350, margin=dict(t=60, b=50, l=20, r=20) )
                fig_dcf_pie.show()
            else: display(Markdown(f"*DCF components zero/negligible. Cannot generate EV Breakdown chart.*"))
        elif fcff_res.get('error'):
             display(Markdown(f"*Cannot generate EV Breakdown chart due to FCFF projection error: {fcff_res.get('error')}*"))
        else:
             display(Markdown(f"*EV calculation failed ({format_currency(total_ev, currency_sym, is_indian=is_indian)}). Cannot generate EV Breakdown chart.*"))
    else: display(Markdown(f"*FCFF results invalid. Cannot generate EV Breakdown chart.*"))

    # --- FCFF Projection Bar Chart ---
    if isinstance(fcff_res, dict) and fcff_res.get('projected_fcffs') and not fcff_res.get('error'):
        years = [f'Y{i+1}' for i in range(len(fcff_res['projected_fcffs']))]; pv_list_clean = [p if np.isfinite(p) else 0 for p in fcff_res.get('pv_fcffs', [])]; fcff_list_clean = [f if np.isfinite(f) else 0 for f in fcff_res.get('projected_fcffs', [])]
        if len(pv_list_clean) == len(years) and len(fcff_list_clean) == len(years):
            df_fcff_plot = pd.DataFrame({'Year': years, 'Forecasted FCFF': fcff_list_clean, 'PV of FCFF': pv_list_clean}); df_fcff_plot['Forecasted FCFF'] = pd.to_numeric(df_fcff_plot['Forecasted FCFF']); df_fcff_plot['PV of FCFF'] = pd.to_numeric(df_fcff_plot['PV of FCFF'])
            df_fcff_melt = df_fcff_plot.melt(id_vars='Year', var_name='Cash Flow Type', value_name='Value')
            fig_fcff_bar = px.bar(df_fcff_melt, x='Year', y='Value', color='Cash Flow Type', title='FCFF Forecast & Present Values', labels={'Value':f'Value ({currency_sym})'}, barmode='group', height=350, color_discrete_map={'Forecasted FCFF':'#5dade2', 'PV of FCFF':'#af7ac5'})
            hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            if is_indian: hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            fig_fcff_bar.update_traces(hovertemplate=hover_bar_format); fig_fcff_bar.update_layout(margin=dict(t=60, b=30, l=50, r=30)); fig_fcff_bar.show()
        else: display(Markdown("*FCFF projection data mismatch/invalid. Cannot plot.*"))
    elif isinstance(fcff_res, dict) and fcff_res.get('error'):
         display(Markdown(f"*Cannot generate FCFF Projection chart due to error: {fcff_res.get('error')}*"))
    else: display(Markdown("*FCFF projection data unavailable or invalid.*"))

    # --- Valuation Conclusion ---
    display(Markdown("### Valuation Conclusion"));
    if isinstance(final_val_dict, dict):
        intrinsic_val = final_val_dict.get('intrinsic_value_per_share', np.nan)
        current_pr_num = current_pr if valid_current_pr else np.nan; conclusion = ""
        if final_val_dict.get('error'): conclusion = f"<span style='color:red'>❌ Valuation Error: {final_val_dict.get('error', '?')}</span>"
        elif valid_intrinsic and intrinsic_val <= 0:
             conclusion = f"<span style='color:orange'>⚠️ Negative DCF Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}.</span> This often results from negative projected free cash flows or high debt relative to negative EV, suggesting the standard DCF model may be inappropriate or requires assumptions beyond historical trends for turnaround."
        elif valid_intrinsic and not np.isfinite(current_pr_num):
             conclusion = f"✅ DCF Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}. Cannot compare (Current Price N/A)."
        elif valid_intrinsic and valid_current_pr : # Standard comparison
             diff_pct = (intrinsic_val / current_pr_num) - 1
             current_fmt = format_currency(current_pr_num, currency_sym, is_indian=is_indian)
             intrinsic_fmt = format_currency(intrinsic_val, currency_sym, is_indian=is_indian)
             if abs(diff_pct) < 0.10: conclusion = f"<span style='color:grey'>~ Fairly Valued:</span> Price ({current_fmt}) vs DCF ({intrinsic_fmt}). Diff: {diff_pct:+.1%}"
             elif diff_pct < 0: overval_pct = (current_pr_num / intrinsic_val) - 1; conclusion = f"<span style='color:#e74c3c'>🔻 Potentially Overvalued:</span> Price ({current_fmt}) is **{overval_pct:+.1%}** above DCF ({intrinsic_fmt})."
             else: conclusion = f"<span style='color:#2ecc71'>✅ Potentially Undervalued:</span> Price ({current_fmt}) suggests **{diff_pct:+.1%}** upside to DCF ({intrinsic_fmt})."
        else: # Fallback if intrinsic value itself is invalid (e.g., NaN)
             conclusion = f"<span style='color:red'>❌ Valuation Error: Intrinsic Value calculation failed.</span>"
    else:
        conclusion = f"<span style='color:red'>❌ Valuation Error: Final valuation data structure invalid.</span>"
    display(Markdown(f"> **{conclusion}**"))

# --- Execute Summary Dashboard Generation ---
if isinstance(final_valuation, dict) and isinstance(fcff_results, dict) and isinstance(stock_data.get('info'), dict) and isinstance(input_sources, dict) and not final_valuation.get('error'):
    try:
        display_summary_dashboard(
            company_name, ticker_symbol, current_price, final_valuation.get('intrinsic_value_per_share'),
            discount_rate, fcff_results.get('terminal_growth_rate'), fcff_results.get('initial_growth_rate'),
            fcff_results.get('growth_source', 'N/A'), final_valuation, stock_data['history'], fcff_results,
            currency_symbol, stock_data['info'], input_sources, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Dashboard Generation: {e}**"))
else:
    display(Markdown("❌ **Skipping Summary Dashboard due to prior errors or invalid data structures.**"))

# %% [markdown]
# ## 10. Assumptions & Disclaimer (Reflects Dynamic/User Inputs & Fallbacks - Growth Logic Updated)

# %%
# Dynamically generate assumption strings based on final values
rf_source_disp = input_sources.get('Rf', '?')
rm_source_disp = input_sources.get('Rm','?')
beta_source_disp = input_sources.get('Beta','?')
c_spread_source_disp = input_sources.get('Credit Spread','N/A')
bench_source_disp = input_sources.get('Benchmarks','?')

# Safely get values from dictionaries
growth_source_disp = fcff_results.get('growth_source', 'N/A') if isinstance(fcff_results, dict) else 'Error/Not Run'
tax_rate_source_disp = wacc_components.get('tax_rate_source', 'N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
equity_source_disp = wacc_components.get('equity_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
weights_source_disp = wacc_components.get('weights_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
terminal_growth_final = fcff_results.get('terminal_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan
initial_growth_final = fcff_results.get('initial_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan
fcff_basis_method_disp = fcff_results.get('fcff_basis_method', 'N/A') if isinstance(fcff_results, dict) else 'Error/Not Run'


# %%
# ----> UPDATED TEXT TO REFLECT NEW GROWTH LOGIC (V21) <----
display(Markdown(f"""
**Key Assumptions & Input Sources:**
*   **Financial Data:** Relies on data provided by Yahoo Finance (`yfinance`). Accuracy subject to provider. Missing data uses fallbacks or causes errors.
*   **Risk-Free Rate (Rf):** Attempted dynamic fetch, **confirmed/overridden by user.** (`Source: {rf_source_disp}` - Default suggestion: `{DEFAULT_RF_RATE_SUGGESTION:.1f}%`)
*   **Expected Market Return (Rm):** Based on **user input.** (`Source: {rm_source_disp}`)
*   **Market Risk Premium (MRP):** Derived from user `Rm` and final `Rf`.
*   **Beta (β):** Dynamically fetched if valid, otherwise **required user input.** (`Source: {beta_source_disp}`)
*   **Cost of Debt (Kd):** Implied if possible & within bounds (`{format_percent(MIN_COST_OF_DEBT)}` - `{format_percent(MAX_COST_OF_DEBT)}`). Otherwise, defaults to `Rf + Credit Spread`.
*   **Credit Spread:** Used only if default Kd triggered. Based on **user input** (default suggestion: `{format_percent(DEFAULT_CREDIT_SPREAD)}`). (`Source: {c_spread_source_disp}`)
*   **FCFF Calculation (T-0 Basis for Projection):** Determined using a **strict positive-average priority sequence:** 1) Avg(`EBITDA-TaxProv-WC-Capex`), 2) Avg(NOPAT method), 3) Avg(CFO+IntAdj method), 4) Avg(CFO-Capex method), 5) Avg(EBITDA-TaxProv), 6) Avg(EBITDAMargin)*RevT0. The first method yielding a positive 3-year average is used as the starting point (`fcff_start_basis`) for projections. Method used: `{fcff_basis_method_disp}`. **Projection halts if no method yields a positive average.** Historical FCFFs for CAGR calculation use standard single-year fallbacks (NOPAT -> CFO+ -> CFO-).
*   **Growth Rates:**
    *   **Initial Growth:** `{format_percent(initial_growth_final)}` used for Year 1, determined *independently* of the T-0 projection basis: (`Source: {growth_source_disp}`)
        1.  Fetch last 3 years (4 data points) of Revenue, FCFF (individual best estimate), and Stock Price (daily).
        2.  **Validate Data:** Check if T-0 (end) is finite and T-3 (start) is positive and finite for each metric.
        3.  Calculate **3-Year CAGR** *only* for metrics with valid start/end data (using `calculate_cagr_modified`). Stock CAGR uses daily price history, preferring 'Adj Close', falling back to 'Close'.
        4.  Calculate the **Geometric Mean** (`geometric_mean_modified`) of the successfully calculated CAGRs from step 3.
        5.  If GM calculation fails or *no valid CAGRs were computed* in step 3, use **Default (`{format_percent(DEFAULT_GROWTH_RATE)}`)**.
        6.  Apply **Bounds:** The result from step 4 or 5 is capped between **`{format_percent(MIN_INITIAL_GROWTH_FLOOR)}` (Floor)** and **`{format_percent(MAX_INITIAL_GROWTH_CEILING)}` (Ceiling)**.
    *   **Terminal Growth (g):** `{format_percent(terminal_growth_final)}` = min(GDP cap=`{format_percent(TERMINAL_GROWTH_RATE_CAP)}`, WACC*`0.67`, Final Initial Growth, Rf*`0.95`), floored at `{format_percent(MIN_GROWTH)}`, ensured `< WACC`. Calculated only if projection proceeds.
*   **Tax Rate:** Effective rate `{tax_rate_source_disp}` (bounded `{format_percent(MIN_TAX_RATE)}` - `{format_percent(MAX_TAX_RATE)}`) or default (`{format_percent(DEFAULT_TAX_RATE)}`).
*   **Capital Structure:** Assumed stable using `{equity_source_disp}` and calculated debt. (`Weights Source: {weights_source_disp}`)
*   **Industry Benchmarks:** Uses **user input** or **generic fallbacks** if input left blank. (`Source: {bench_source_disp}`)
*   **Currency Formatting:** Uses ₹/Lakh/Crore for `.NS` tickers, standard K/M/B/T otherwise.

**Limitations:**
*   **Model Sensitivity:** DCF highly sensitive to inputs. **Sensitivity analysis (not included) crucial.**
*   **Input Quality:** Validity depends on user inputs and appropriateness of fallbacks.
*   **Historical Data Availability:** Financial statement data (needed for Rev/FCFF CAGR) is often limited to ~4 years via `yfinance`. Accuracy of 3-year CAGR depends heavily on these few points. Stock price history is usually more extensive but relies on finding suitable start/end dates. CAGR calculation is skipped if start/end data is invalid or start is non-positive.
*   **FCFF Calculation & Projection:** FCFF calculation for the **projection start point** relies on finding a method that yields a *positive average* over recent years. If the company consistently has negative calculated FCFF across multiple methods/averages, the model **cannot project FCFF** and the DCF valuation will fail. Averaging methods smooth results and may obscure underlying issues. The FCFF used for CAGR calculation (individual best estimate) may differ from the average used as the projection basis.
*   **CAGR/GM Modifications:** The modified CAGR handles non-positive end values (relative to positive start) but requires a positive start value. The modified GM handles negative rates but falls back to arithmetic mean if rates are <= -100% or other errors occur. GM is less meaningful with highly disparate or negative rates.
*   **Dynamic Data Reliability:** Depends on `yfinance` accuracy/availability. Rf proxies need verification. Stock history column names ('Adj Close', 'Close') assumed.
*   **Simplifications:** Linear decay, constant rates/structure. Ignores qualitative factors, macro shocks, sentiment. Some FCFF calculation methods (User, Avg Ebitda-Tax, Avg Margin) are simplifications.
*   **Fallback Accuracy:** Default growth rate (6.5%) and generic benchmarks may not be representative.
*   **Data Completeness:** Performance degrades with missing data. *Cross-verify key data.*

**Disclaimer:**
*For **educational and illustrative purposes only**. **NOT investment advice**. Models have limitations. Users MUST conduct own due diligence and consult a qualified financial advisor before investing. Creators NOT liable for actions based on this report.*
"""))




# %%
````

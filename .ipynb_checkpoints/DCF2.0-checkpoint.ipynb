````python
# %% [markdown]
# # Dynamic Stock Financial Analysis & DCF Valuation
# **(Interactive Report - Dynamic/User Input Driven Assumptions - V21.2 - User FCFF Choice & Fixes)**
#
# **Analyst:** Simulated AI Research Analyst
#
# **Objective:** To perform a comprehensive financial analysis and estimate the intrinsic value per share using DCF. Prioritizes dynamic data, user input, and defaults. **Calculates potential positive average FCFF bases (4yr avg), presents successful options to user for selection. Initial growth still uses GM of individual 3yr CAGRs. Ratio plots fixed.**
#
# **Methodology:**
# 1.  **User Input & Dynamic Fetching:** Standard setup.
# 2.  **Data Acquisition:** Standard setup, including 4y daily history.
# 3.  **WACC Calculation:** Standard setup.
# 4.  **FCFF Calculation & Projection:**
#     *   Calculate historical FCFF (T-0 to T-3) individually using standard fallbacks (`_calculate_fcff_single_year_best_estimate` with verbose logging) for CAGR calculation. **(Logging clarifies this is for CAGR only)**.
#     *   **Determine Potential Projection Bases:** Use new helper `_calculate_all_fcff_bases_v21_4yr` to calculate 4yr averages for all standard methods.
#     *   **User Selection:** If positive bases found, present options (Method, Formula, Avg Value) to user for selection.
#     *   **Fallback/Halt:** If no positive average found, try T-0 individual; if still negative/invalid, halt projection.
#     *   **Estimate initial growth:** (Uses individual historical FCFFs for FCFF CAGR) Calculate 3yr CAGRs for Rev, FCFF, Stock. Calculate GM. Fallback/Bound.
#     *   **Project FCFF (Conditional):** Project forward from *user-selected* or fallback `fcff_start_basis` using decaying growth rate.
# 5.  **Terminal Value:** Standard calculation.
# 6.  **Discounting & Valuation:** Standard calculation.
# 7.  **Ratio Analysis:** Standard calculation. **(Plotting logic corrected)**.
# 8.  **Interactive Visualization:** Standard setup.

# %% [markdown]
# ## 1. Setup and Libraries

# %%
# Standard Libraries
import warnings
import sys
import numpy as np
import pandas as pd
import datetime
import math # For isnan, isinf checks
import traceback # For detailed error logging

# Data Fetching
import yfinance as yf

# Visualization (Using Plotly)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

# Display Utilities
from IPython.display import display, Markdown, HTML

# Settings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pio.templates.default = "plotly_white" # Cleaner Plotly template

# %% [markdown]
# ## 2. Configuration and Helper Functions (FCFF Helpers Updated V21.2)

# %%
# --- Model Configuration ---
DEFAULT_TICKER = "MUTHOOTFIN.NS"
FORECAST_YEARS = 5
TERMINAL_GROWTH_RATE_CAP = 0.05
DEFAULT_GROWTH_RATE = 0.065 # Used if CAGR/GM fails
DEFAULT_RF_RATE_SUGGESTION = 7.0
DEFAULT_CURRENCY_SYMBOL = "₹"
DEFAULT_TAX_RATE = 0.25
DEFAULT_CREDIT_SPREAD = 0.015 # 1.5%
DEFAULT_MARKET_RISK_PREMIUM = 0.06 # 6.0%

# --- Generic Fallback Benchmark Ratios ---
# IMPORTANT: Store percentages as numbers, e.g., 19.05 for 19.05%
DEFAULT_BENCHMARK_RATIOS = {
    'P/E': 17.3, 'P/B': 2.5, 'EV/EBITDA': 11.0, 'ROE': 19.05,
    'ROA': 3.92, 'ROCE': 11.85,'Asset Turnover': 0.15, # Note: Asset Turnover is not a percentage
    'Current Ratio': 1.5, 'Quick Ratio': 1.5
}

# --- Model Stability Bounds ---
MIN_WACC = 0.06; MAX_WACC = 0.25; MIN_TAX_RATE = 0.10; MAX_TAX_RATE = 0.40
MIN_GROWTH = 0.00; MIN_INITIAL_GROWTH_FLOOR = 0.011; MAX_INITIAL_GROWTH_CEILING = 0.15
MIN_COST_OF_DEBT = 0.03; MAX_COST_OF_DEBT = 0.18

# --- Ratio Keys Needed for Benchmarking ---
REQUIRED_BENCHMARK_RATIOS = list(DEFAULT_BENCHMARK_RATIOS.keys())

# --- Helper Functions ---
# Function to print statement keys (unchanged)
def print_keys(statement, name):
    print(f"\n--- Available Keys in {name} ---")
    if isinstance(statement, pd.DataFrame):
        valid_indices = [idx for idx in statement.index if idx is not None]
        print(", ".join(map(str, valid_indices)))
    elif isinstance(statement, dict): print(", ".join(statement.keys()))
    print("------------------------------------\n")

# safe_get function (unchanged)
def safe_get(data, key_list, statement_name="Statement", default=np.nan): # Default to NaN
    if not isinstance(key_list, list): key_list = [key_list]
    value = None
    found_key = None # Track which key was successful
    for key in key_list:
        current_value = None; raw_val = None
        data_source = None
        if isinstance(data, pd.Series):
            data_source = f"'{statement_name}' Series"
            if key in data.index:
                try:
                    raw_val = data.loc[key]
                    if isinstance(raw_val, (int, float)) and np.isfinite(raw_val): current_value = float(raw_val)
                    elif isinstance(raw_val, pd.Series):
                        if not raw_val.empty and pd.notna(raw_val.iloc[0]) and np.isfinite(raw_val.iloc[0]): current_value = float(raw_val.iloc[0])
                    elif pd.notna(raw_val):
                        try_val = float(raw_val);
                        if np.isfinite(try_val): current_value = try_val
                except (KeyError, TypeError, IndexError, ValueError): pass
        elif isinstance(data, dict):
            data_source = f"'{statement_name}' Dict"
            raw_val = data.get(key)
            if raw_val is not None:
                try:
                    # Handle common non-numeric strings
                    if isinstance(raw_val, str) and raw_val.lower() in ['n/a', 'none', '', '--', '-']: pass
                    else: temp_val = float(raw_val);
                    if np.isfinite(temp_val): current_value = temp_val
                except (TypeError, ValueError): pass
        if current_value is not None:
            value = current_value;
            found_key = key
            # print(f"    ✅ Found '{found_key}' in {data_source}: {value}") # DEBUG Line
            break # Stop searching once a valid key is found

    # if value is None: print(f"    ❌ Could not find '{key_list}' in {data_source}. Using default: {default}") # DEBUG Line
    if value is None: value = default

    return value

# format_currency (unchanged)
def format_currency(value, symbol="₹", precision=2, is_indian=False):
    if value is None or pd.isna(value) or not math.isfinite(value): return "N/A"
    if abs(value) > 1e18: return "Inf" # Handle potential overflows
    sign = "-" if value < 0 else ""; abs_val = abs(value)
    if is_indian:
        symbol = "₹"
        if abs_val >= 1e7: formatted_val = f"{abs_val / 1e7:.{precision}f} Cr"
        elif abs_val >= 1e5: formatted_val = f"{abs_val / 1e5:.{precision}f} Lakh"
        else: formatted_val = f"{abs_val:,.{precision}f}"
        return f"{sign}{symbol} {formatted_val}"
    else: # Standard K/M/B/T formatting
        if abs_val >= 1e12: formatted_val = f"{abs_val / 1e12:.{precision}f}T"
        elif abs_val >= 1e9: formatted_val = f"{abs_val / 1e9:.{precision}f}B"
        elif abs_val >= 1e6: formatted_val = f"{abs_val / 1e6:.{precision}f}M"
        elif abs_val >= 1e3: formatted_val = f"{abs_val / 1e3:.{precision}f}K"
        else: formatted_val = f"{abs_val:.{precision}f}"
        return f"{sign}{symbol}{formatted_val}"

# format_percent (unchanged)
def format_percent(value, precision=1):
     if value is None or pd.isna(value) or not math.isfinite(value): return "N/A"
     if abs(value) > 1e5: return "Inf" # Handle large percentages
     return f"{value * 100:.{precision}f}%"

# get_validated_float_input (modified slightly for ratio input clarity)
def get_validated_float_input(prompt_message, default_value=None, allow_blank=False, min_val=None, max_val=None, is_percentage=False):
    while True:
        prompt_with_default = prompt_message
        if is_percentage: prompt_with_default += " (e.g., enter '15' for 15%)"
        if default_value is not None: prompt_with_default += f" [Default: {default_value}]: "
        elif allow_blank: prompt_with_default += f" [Leave blank to skip/use fallback]: "
        else: prompt_with_default += ": "
        try:
            user_input_str = input(prompt_with_default).strip()
            if not user_input_str:
                if allow_blank: return None
                if default_value is not None:
                    try: return float(default_value)
                    except ValueError: raise ValueError(f"Internal Error: Invalid default value '{default_value}'.")
                raise ValueError("Input cannot be empty.")
            # Allow '%' sign but remove it for conversion
            if user_input_str.endswith('%'): user_input_str = user_input_str[:-1]
            value = float(user_input_str)
            if min_val is not None and value < min_val: raise ValueError(f"Value must be >= {min_val}.")
            if max_val is not None and value > max_val: raise ValueError(f"Value must be <= {max_val}.")
            # If it's meant to be a percentage decimal (like Rf), divide by 100
            # if is_percentage_decimal: value /= 100.0
            return value
        except ValueError as e: print(f"❌ Invalid input: {e}. Please enter a valid number.")

# try_get_dynamic_rf (unchanged)
def try_get_dynamic_rf(rf_ticker_symbol="^TNX"):
    print(f"Attempting to fetch dynamic Risk-Free Rate ({rf_ticker_symbol})...")
    try:
        rf_ticker = yf.Ticker(rf_ticker_symbol); hist = rf_ticker.history(period="5d")
        if not hist.empty and 'Close' in hist.columns:
            last_close = hist['Close'].iloc[-1]
            if pd.notna(last_close):
                dynamic_rf = last_close / 100.0 # Assume fetched value is percentage
                if 0.001 < dynamic_rf < 0.20: # Check plausible range 0.1% to 20%
                     print(f"✅ Dynamically fetched Rf ({rf_ticker_symbol}): {format_percent(dynamic_rf)}"); return dynamic_rf
                else: print(f"⚠️ Fetched Rf value ({format_percent(dynamic_rf)}) out of typical range (0.1%-20%). Ignoring."); return None
            else: print(f"⚠️ Fetched Rf data was invalid (NaN) for {rf_ticker_symbol}."); return None
        else: print(f"⚠️ Could not fetch recent data for {rf_ticker_symbol}."); return None
    except Exception as e: print(f"❌ Error fetching dynamic Rf ({rf_ticker_symbol}): {e}"); return None

# --- Growth Calculation Helpers (Unchanged) ---
def calculate_cagr_modified(start_val, end_val, num_years):
    if pd.isna(start_val) or pd.isna(end_val) or not math.isfinite(start_val) or not math.isfinite(end_val) or num_years <= 0: return np.nan
    if start_val <= 1e-9: return np.nan # Start value must be positive
    cagr = np.nan
    try:
        # Handle negative end values relative to positive start
        if end_val <= 1e-9:
            end_val_abs = max(abs(end_val), 1e-9); start_val_abs = abs(start_val)
            ratio = end_val_abs / start_val_abs
            if ratio > 0: cagr_magnitude = ratio ** (1 / num_years) - 1; cagr = -cagr_magnitude # Return negative growth magnitude
        else: # Standard positive start and end
            ratio = end_val / start_val
            if ratio > 0: cagr = ratio ** (1 / num_years) - 1
    except (ValueError, OverflowError, ZeroDivisionError) as e: cagr = np.nan # Catch math errors
    # Check for unreasonable CAGR results (e.g., >2000% or < -99.9%)
    if pd.notna(cagr) and not (-0.999 < cagr < 20.0): cagr = np.nan
    return cagr

def geometric_mean_modified(rates):
    valid_rates = [r for r in rates if pd.notna(r) and math.isfinite(r)]
    if not valid_rates: return np.nan
    if len(valid_rates) == 1: return valid_rates[0]
    offset = 1.0; factors_offset = [r + offset for r in valid_rates]
    # Check if any rate is <= -100% (factor <= 0)
    if any(f <= 1e-9 for f in factors_offset):
         print("⚠️ Geometric mean calculation failed due to rate <= -100%. Falling back to arithmetic mean.")
         return np.mean(valid_rates)
    n = len(valid_rates); product = 1.0
    try:
        for factor in factors_offset: product *= factor
        # Check for negative product with even root (imaginary number)
        if product < 0 and n % 2 == 0:
             print("⚠️ Geometric mean calculation failed (negative base for even root). Falling back to arithmetic mean.")
             return np.mean(valid_rates)
        # Calculate GM using offset method
        gm_offset = np.sign(product) * (np.abs(product) ** (1/n)); gm = gm_offset - offset
        # Check for unreasonable GM results
        if pd.notna(gm) and not (-0.999 < gm < 20.0):
            print(f"⚠️ Geometric mean result ({format_percent(gm)}) out of reasonable bounds. Falling back to arithmetic mean.")
            return np.mean(valid_rates)
        return gm
    except (ValueError, OverflowError, ZeroDivisionError, TypeError) as e:
        print(f"⚠️ Geometric mean calculation encountered numeric error ({e}). Falling back to arithmetic mean.")
        return np.mean(valid_rates)

# --- FCFF Component Fetcher (Added Verbose Logging V21.2) ---
def _get_fcff_components(fin_col, bs_col, cf_col, bs_prev_col, log_list):
    """Fetches all components needed for various FCFF calculations for a single year with logging."""
    components = {}
    log_list.append("--- Fetching FCFF Components ---")

    # Revenue
    rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
    components['revenue'] = safe_get(fin_col, rev_keys, "Income Stmt", default=np.nan)
    if np.isfinite(components['revenue']): log_list.append(f"    ✅ Found '{rev_keys[0]}' in Income Stmt: {format_currency(components['revenue'], '', 0)}") # Assume first key is primary
    else: log_list.append(f"    ❌ Could not find Revenue in Income Stmt. Tried: {', '.join(rev_keys)}. Using default: nan")

    # EBIT
    ebit_keys = ["EBIT", "Ebit", "Operating Income"]
    # Add more potential keys if needed
    ebit_alt_keys = ["Total Operating Income As Reported", "Earnings Before Interest and Tax"]
    all_ebit_keys = ebit_keys + ebit_alt_keys
    components['ebit'] = safe_get(fin_col, all_ebit_keys, "Income Stmt", default=np.nan)
    if np.isfinite(components['ebit']): log_list.append(f"    ✅ Found EBIT (or alternative) in Income Stmt: {format_currency(components['ebit'], '', 0)}")
    else:
        log_list.append(f"    ❌ Could not find EBIT (or alternatives) in Income Stmt. Tried: {', '.join(all_ebit_keys)}. Using default: nan")
        # Fallback: EBIT = Pretax Income + Interest Expense
        log_list.append("    -> EBIT not found directly. Trying fallback: Pretax Income + Interest Expense...")
        pretax_inc_keys = ['Pretax Income', 'EBT', 'Earnings Before Tax']
        pretax_inc = safe_get(fin_col, pretax_inc_keys, "Income Stmt (for EBIT fallback)", default=np.nan)
        interest_exp_keys = ['Interest Expense', 'Interest Expense Net']
        int_exp_for_ebit = abs(safe_get(fin_col, interest_exp_keys, "Income Stmt (for EBIT fallback)", default=np.nan)) # Use absolute value
        if np.isfinite(pretax_inc): log_list.append(f"    ✅ Found '{pretax_inc_keys[0]}' in Income Stmt (for EBIT fallback): {format_currency(pretax_inc, '', 0)}")
        else: log_list.append(f"    ❌ Could not find Pretax Income for EBIT fallback.")
        if np.isfinite(int_exp_for_ebit): log_list.append(f"    ✅ Found '{interest_exp_keys[0]}' in Income Stmt (for EBIT fallback): {format_currency(int_exp_for_ebit, '', 0)}")
        else: log_list.append(f"    ❌ Could not find Interest Expense for EBIT fallback.")

        if np.isfinite(pretax_inc) and np.isfinite(int_exp_for_ebit):
            components['ebit'] = pretax_inc + int_exp_for_ebit
            log_list.append(f"    ✅ Derived EBIT = {format_currency(pretax_inc,'',0)} (Pretax) + {format_currency(int_exp_for_ebit,'',0)} (Interest) = {format_currency(components['ebit'],'',0)}")
        else:
            log_list.append(f"    ❌ Failed to derive EBIT from fallback.")
            components['ebit'] = np.nan # Ensure it remains NaN


    # Depreciation & Amortization
    dep_amort_keys = ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"]
    components['dep_amort'] = safe_get(cf_col, dep_amort_keys, "Cash Flow", default=np.nan)
    if not np.isfinite(components['dep_amort']):
         components['dep_amort'] = safe_get(fin_col, ["Depreciation And Amortization", "Depreciation Income Statement"], "Income Stmt (Dep Fallback)", default=np.nan)
         if np.isfinite(components['dep_amort']): log_list.append(f"    ⚠️ Found Dep&Amort in Income Stmt (CF failed): {format_currency(components['dep_amort'], '', 0)}")
         else: log_list.append(f"    ❌ Could not find Dep&Amort in CF or IS. Tried: {', '.join(dep_amort_keys)}. Using default: nan")
    else:
        log_list.append(f"    ✅ Found '{dep_amort_keys[0]}' in Cash Flow: {format_currency(components['dep_amort'], '', 0)}")

    # EBITDA (Calculated)
    if np.isfinite(components['ebit']) and np.isfinite(components['dep_amort']):
        components['ebitda'] = components['ebit'] + components['dep_amort']
        log_list.append(f"    Computed EBITDA = {format_currency(components['ebit'],'',0)} (EBIT) + {format_currency(components['dep_amort'],'',0)} (Dep&Amort) = {format_currency(components['ebitda'],'',0)}")
    else:
        components['ebitda'] = np.nan
        log_list.append(f"    ⚠️ Cannot compute EBITDA (EBIT={components['ebit']}, Dep&Amort={components['dep_amort']})")


    # CapEx
    capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
    capex_val = safe_get(cf_col, capex_keys, "Cash Flow", default=np.nan)
    components['capex'] = abs(capex_val) if np.isfinite(capex_val) else np.nan # Use absolute value
    if np.isfinite(components['capex']): log_list.append(f"    ✅ Found '{capex_keys[0]}' in Cash Flow: {format_currency(capex_val, '', 0)}")
    if np.isfinite(capex_val) and capex_val != components['capex']: log_list.append(f"    -> Using CapEx = {format_currency(components['capex'],'',0)} (absolute value of found item)")
    elif not np.isfinite(components['capex']): log_list.append(f"    ❌ Could not find CapEx in Cash Flow. Tried: {', '.join(capex_keys)}. Using default: nan")


    # Change in Working Capital (WC) - Prioritize CF, fallback to BS calc
    wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"] # Add more if needed
    components['wc_cf'] = safe_get(cf_col, wc_keys, "Cash Flow (Direct WC)", default=np.nan)
    if np.isfinite(components['wc_cf']):
        components['wc'] = components['wc_cf'] # Use CF value directly
        components['wc_src'] = "CF Direct"
        log_list.append(f"    ✅ Found '{wc_keys[0]}' in Cash Flow (Direct WC): {format_currency(components['wc'], '', 0)}")
        log_list.append(f"    ✅ Using WC from {components['wc_src']}: {format_currency(components['wc'],'',0)}")
    else:
        log_list.append(f"    ℹ️ Direct WC not found in CF Statement. Trying BS calculation...")
        components['wc_src'] = "BS Calc"
        components['wc'] = np.nan # Reset wc to NaN before trying BS calc
        # Check if necessary balance sheet columns are present
        if isinstance(bs_prev_col, pd.Series) and isinstance(bs_col, pd.Series):
             curr_assets_keys = ['Current Assets', 'Total Current Assets']
             curr_liab_keys = ['Current Liabilities', 'Total Current Liabilities']
             curr_assets = safe_get(bs_col, curr_assets_keys, 'BS Current', default=np.nan)
             curr_liab = safe_get(bs_col, curr_liab_keys, 'BS Current', default=np.nan)
             prev_assets = safe_get(bs_prev_col, curr_assets_keys, 'BS Previous', default=np.nan)
             prev_liab = safe_get(bs_prev_col, curr_liab_keys, 'BS Previous', default=np.nan)

             if np.isfinite(curr_assets) and np.isfinite(curr_liab): curr_wc_val = curr_assets - curr_liab
             else: curr_wc_val = np.nan; log_list.append("      ❌ Missing Current Assets or Liabilities in Current BS.")
             if np.isfinite(prev_assets) and np.isfinite(prev_liab): prev_wc_val = prev_assets - prev_liab
             else: prev_wc_val = np.nan; log_list.append("      ❌ Missing Current Assets or Liabilities in Previous BS.")

             if np.isfinite(curr_wc_val) and np.isfinite(prev_wc_val):
                 components['wc'] = curr_wc_val - prev_wc_val
                 log_list.append(f"      Current WC (A-L): {format_currency(curr_wc_val,'',0)}")
                 log_list.append(f"      Previous WC (A-L): {format_currency(prev_wc_val,'',0)}")
                 log_list.append(f"    ✅ Calculated WC Change (BS): {format_currency(components['wc'],'',0)}")
                 log_list.append(f"    ✅ Using WC from {components['wc_src']}: {format_currency(components['wc'],'',0)}")

             else: components['wc'] = np.nan; components['wc_src'] = "BS Fail"; log_list.append("    ❌ WC Calculation from BS Failed (Missing components).")
        else: components['wc'] = np.nan; components['wc_src'] = "No Prev BS"; log_list.append("    ❌ WC Calculation from BS Failed (Missing current or previous BS column).")

    # CFO
    cfo_keys = ["Cash Flow From Operating Activities", "Operating Cash Flow", "Total Cash From Operating Activities"]
    components['cfo'] = safe_get(cf_col, cfo_keys, "Cash Flow", default=np.nan)
    if np.isfinite(components['cfo']): log_list.append(f"    ✅ Found '{cfo_keys[0]}' in Cash Flow: {format_currency(components['cfo'], '', 0)}")
    else: log_list.append(f"    ❌ Could not find CFO in Cash Flow. Tried: {', '.join(cfo_keys)}. Using default: nan")

    # Interest Expense (already fetched for EBIT fallback, use that)
    interest_expense_keys = ['Interest Expense', 'Interest Expense Net']
    components['int_exp'] = abs(safe_get(fin_col, interest_expense_keys, "Income Stmt", default=np.nan))
    if np.isfinite(components['int_exp']): log_list.append(f"    ✅ Found '{interest_expense_keys[0]}' in Income Stmt: {format_currency(components['int_exp'], '', 0)}")
    elif np.isfinite(int_exp_for_ebit): # Check if found during EBIT fallback
         components['int_exp'] = int_exp_for_ebit
         log_list.append(f"    ℹ️ Using Interest Expense found during EBIT fallback: {format_currency(components['int_exp'], '', 0)}")
    else: log_list.append(f"    ❌ Could not find Interest Expense in Income Stmt. Using default: nan")


    # Tax Provision
    tax_prov_keys = ['Tax Provision', 'Income Tax Expense']
    components['tax_prov'] = safe_get(fin_col, tax_prov_keys, "Income Stmt", default=np.nan)
    if np.isfinite(components['tax_prov']): log_list.append(f"    ✅ Found '{tax_prov_keys[0]}' in Income Stmt: {format_currency(components['tax_prov'], '', 0)}")
    else: log_list.append(f"    ❌ Could not find Tax Provision in Income Stmt. Tried: {', '.join(tax_prov_keys)}. Using default: nan")

    # Final check on absolute values (already done for CapEx, IntExp)
    # No change needed here

    return components

# --- FCFF Calculation Helper: Individual Year Best Estimate (Added Logging V21.2) ---
def _calculate_fcff_single_year_best_estimate(fin_col, bs_col, cf_col, bs_prev_col, tax_rt, log_list_parent):
    """
    Calculates single year FCFF using NOPAT -> CFO+IntAdj -> CFO-Capex fallbacks.
    Returns best possible finite value (can be negative) or NaN. Method string also returned.
    Includes detailed logging.
    """
    log_list_local = [] # Use local log list for component fetching
    c = _get_fcff_components(fin_col, bs_col, cf_col, bs_prev_col, log_list_local)
    log_list_parent.extend(log_list_local) # Append component logs to parent

    if not np.isfinite(tax_rt): tax_rt = np.nan # Handle invalid tax rate input

    fcff_val = np.nan
    method_used = "All Methods Failed"

    # Method 1: NOPAT = EBIT*(1-T) + Dep - Capex - WC
    log_list_parent.append(f"  FCFF Method Tried: NOPAT = {format_currency(c.get('ebit',np.nan),'',0)}*(1-{format_percent(tax_rt,1)}) + {format_currency(c.get('dep_amort',np.nan),'',0)} - {format_currency(c.get('capex',np.nan),'',0)} - {format_currency(c.get('wc',np.nan),'',0)} (WC Source: {c.get('wc_src','?')})")
    m1_components = [c.get('ebit'), tax_rt, c.get('dep_amort'), c.get('capex'), c.get('wc')]
    if all(map(np.isfinite, m1_components)):
        try:
            nopat_m1 = c['ebit'] * (1 - tax_rt)
            fcff_m1 = nopat_m1 + c['dep_amort'] - c['capex'] - c['wc']
            if np.isfinite(fcff_m1):
                 fcff_val = fcff_m1; method_used = f"NOPAT (WC:{c.get('wc_src','?')})"
                 log_list_parent.append(f"  ---> Result: {format_currency(fcff_val,'',0)} (Using Method: {method_used})")
            else: log_list_parent.append("  ---> NOPAT Calc Result Invalid (NaN/Inf).")
        except Exception as e: log_list_parent.append(f"  ---> NOPAT Calc Error: {e}.")
    else: log_list_parent.append(f"  ---> NOPAT Failed (Missing Components: EBIT={np.isfinite(m1_components[0])}, Tax={np.isfinite(m1_components[1])}, Dep={np.isfinite(m1_components[2])}, Capex={np.isfinite(m1_components[3])}, WC={np.isfinite(m1_components[4])}).")

    # Method 2: CFO + Int*(1-T) - Capex (Only if Method 1 failed)
    if not np.isfinite(fcff_val):
        log_list_parent.append(f"  FCFF Method Tried: CFO+IntAdj = {format_currency(c.get('cfo',np.nan),'',0)} + {format_currency(c.get('int_exp',np.nan),'',0)}*(1-{format_percent(tax_rt,1)}) - {format_currency(c.get('capex',np.nan),'',0)}")
        m2_components = [c.get('cfo'), c.get('capex'), c.get('int_exp'), tax_rt]
        if all(map(np.isfinite, m2_components)):
            try:
                int_tax_shield = c['int_exp'] * (1 - tax_rt) # Correct formula is Int*(1-T) added to CFO-Capex
                fcff_m2 = c['cfo'] - c['capex'] + int_tax_shield
                if np.isfinite(fcff_m2):
                    fcff_val = fcff_m2; method_used = "CFO+IntAdj"
                    log_list_parent.append(f"  ---> Result: {format_currency(fcff_val,'',0)} (Using Method: {method_used})")
                else: log_list_parent.append("  ---> CFO+IntAdj Calc Result Invalid (NaN/Inf).")
            except Exception as e: log_list_parent.append(f"  ---> CFO+IntAdj Calc Error: {e}.")
        else: log_list_parent.append(f"  ---> CFO+IntAdj Failed (Missing Components: CFO={np.isfinite(m2_components[0])}, Capex={np.isfinite(m2_components[1])}, IntExp={np.isfinite(m2_components[2])}, Tax={np.isfinite(m2_components[3])}).")

    # Method 3: CFO - Capex (Only if Methods 1 & 2 failed)
    if not np.isfinite(fcff_val):
        log_list_parent.append(f"  FCFF Method Tried: CFO-Capex = {format_currency(c.get('cfo',np.nan),'',0)} - {format_currency(c.get('capex',np.nan),'',0)}")
        m3_components = [c.get('cfo'), c.get('capex')]
        if all(map(np.isfinite, m3_components)):
            try:
                fcff_m3 = c['cfo'] - c['capex']
                if np.isfinite(fcff_m3):
                    fcff_val = fcff_m3; method_used = "CFO-CapEx"
                    log_list_parent.append(f"  ---> Result: {format_currency(fcff_val,'',0)} (Using Method: {method_used})")
                else: log_list_parent.append("  ---> CFO-Capex Calc Result Invalid (NaN/Inf).")
            except Exception as e: log_list_parent.append(f"  ---> CFO-Capex Calc Error: {e}.")
        else: log_list_parent.append(f"  ---> CFO-Capex Failed (Missing Components: CFO={np.isfinite(m3_components[0])}, Capex={np.isfinite(m3_components[1])}).")

    if not np.isfinite(fcff_val):
         log_list_parent.append(f"  ---> Result: N/A (All FCFF methods failed)")

    return fcff_val, method_used

# --- FCFF Basis Calculation Helper (V21.2 - Calculates ALL positive bases) ---
def _calculate_all_fcff_bases_v21_4yr(financials_df, balance_sheet_df, cashflow_df, tax_rt, log_list_parent):
    """
    Calculates FCFF using multiple averaging methods over up to 4 years.
    Returns a dictionary of {method_name: {'avg_fcff': value, 'valid_years': count, 'formula': str}}
    for methods that yield a POSITIVE average.
    Includes detailed logging.
    """
    positive_bases = {} # Store successful methods and their results
    max_avg_years = 4 # Use last 4 years: T-0, T-1, T-2, T-3
    num_statement_years = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    log_list_parent.append(f"--- Calculating Potential FCFF Projection Bases (Positive Average Priority V21.2 - {max_avg_years}yr Target) ---")

    # Check available data years
    years_to_average = min(num_statement_years, max_avg_years)
    if years_to_average < 1:
        log_list_parent.append(f"  ❌ Less than 1 year of data available. Cannot perform averaging.")
        return positive_bases # Return empty dict

    # Check for sufficient BS data needed for WC calcs across the averaging period
    bs_sufficient_for_wc = balance_sheet_df.shape[1] >= (years_to_average + 1)
    if not bs_sufficient_for_wc:
        log_list_parent.append(f"  ⚠️ WARNING: Insufficient Balance Sheet data ({balance_sheet_df.shape[1]} yrs) for full WC calc over {years_to_average} avg years (need {years_to_average + 1}). Methods relying on BS-based WC may fail or use CF fallback.")

    # --- Define Calculation Functions for Each Method (Single Year - with logging) ---
    def calc_method0(fin, bs, cf, bs_prev, log_list_calc): # EBITDA - TaxProv - WC - CapEx
        c = _get_fcff_components(fin, bs, cf, bs_prev, log_list_calc)
        if all(map(np.isfinite, [c.get('ebitda'), c.get('tax_prov'), c.get('wc'), c.get('capex')])):
            res = c['ebitda'] - c['tax_prov'] - c['wc'] - c['capex']
            log_list_calc.append(f"    Method 0 Calc: {format_currency(c['ebitda'],'',0)} - {format_currency(c['tax_prov'],'',0)} - {format_currency(c['wc'],'',0)} - {format_currency(c['capex'],'',0)} = {format_currency(res,'',0)}")
            return res
        log_list_calc.append(f"    Method 0 Failed (Components: EBITDA={np.isfinite(c.get('ebitda'))}, TaxProv={np.isfinite(c.get('tax_prov'))}, WC={np.isfinite(c.get('wc'))}, Capex={np.isfinite(c.get('capex'))})")
        return np.nan

    def calc_method1(fin, bs, cf, bs_prev, tax_rate, log_list_calc): # NOPAT = EBIT*(1-T)+D&A-Capex-WC
        c = _get_fcff_components(fin, bs, cf, bs_prev, log_list_calc)
        if all(map(np.isfinite, [c.get('ebit'), tax_rate, c.get('dep_amort'), c.get('capex'), c.get('wc')])):
            nopat = c['ebit'] * (1 - tax_rate)
            res = nopat + c['dep_amort'] - c['capex'] - c['wc']
            log_list_calc.append(f"    Method 1 Calc (NOPAT): ({format_currency(c['ebit'],'',0)}*(1-{format_percent(tax_rate,1)})) + {format_currency(c['dep_amort'],'',0)} - {format_currency(c['capex'],'',0)} - {format_currency(c['wc'],'',0)} = {format_currency(res,'',0)}")
            return res
        log_list_calc.append(f"    Method 1 Failed (Components: EBIT={np.isfinite(c.get('ebit'))}, Tax={np.isfinite(tax_rate)}, Dep={np.isfinite(c.get('dep_amort'))}, Capex={np.isfinite(c.get('capex'))}, WC={np.isfinite(c.get('wc'))})")
        return np.nan

    def calc_method2(fin, bs, cf, bs_prev, tax_rate, log_list_calc): # CFO + Int*(1-T) - Capex
        c = _get_fcff_components(fin, bs, cf, bs_prev, log_list_calc)
        if all(map(np.isfinite, [c.get('cfo'), c.get('capex'), c.get('int_exp'), tax_rate])):
            int_tax_shield = c['int_exp'] * (1 - tax_rate)
            res = c['cfo'] - c['capex'] + int_tax_shield
            log_list_calc.append(f"    Method 2 Calc (CFO+IntAdj): {format_currency(c['cfo'],'',0)} - {format_currency(c['capex'],'',0)} + ({format_currency(c['int_exp'],'',0)}*(1-{format_percent(tax_rate,1)})) = {format_currency(res,'',0)}")
            return res
        log_list_calc.append(f"    Method 2 Failed (Components: CFO={np.isfinite(c.get('cfo'))}, Capex={np.isfinite(c.get('capex'))}, IntExp={np.isfinite(c.get('int_exp'))}, Tax={np.isfinite(tax_rate)})")
        return np.nan

    def calc_method3(fin, bs, cf, bs_prev, log_list_calc): # CFO - CapEx
        c = _get_fcff_components(fin, bs, cf, bs_prev, log_list_calc)
        if all(map(np.isfinite, [c.get('cfo'), c.get('capex')])):
            res = c['cfo'] - c['capex']
            log_list_calc.append(f"    Method 3 Calc (CFO-Capex): {format_currency(c['cfo'],'',0)} - {format_currency(c['capex'],'',0)} = {format_currency(res,'',0)}")
            return res
        log_list_calc.append(f"    Method 3 Failed (Components: CFO={np.isfinite(c.get('cfo'))}, Capex={np.isfinite(c.get('capex'))})")
        return np.nan

    # Definition only, no calculation here
    def calc_method4(fin, bs, cf, bs_prev, log_list_calc): # EBITDA - TaxProv (Proxy Component)
        c = _get_fcff_components(fin, bs, cf, bs_prev, log_list_calc)
        if all(map(np.isfinite, [c.get('ebitda'), c.get('tax_prov')])):
            res = c['ebitda'] - c['tax_prov']
            log_list_calc.append(f"    Method 4 Component Calc (EBITDA-TaxProv): {format_currency(c['ebitda'],'',0)} - {format_currency(c['tax_prov'],'',0)} = {format_currency(res,'',0)}")
            return res
        log_list_calc.append(f"    Method 4 Component Failed (Components: EBITDA={np.isfinite(c.get('ebitda'))}, TaxProv={np.isfinite(c.get('tax_prov'))})")
        return np.nan

    # Definition only, no calculation here
    def calc_method5(fin, bs, cf, bs_prev, log_list_calc): # EBITDA Margin (Proxy Component)
        c = _get_fcff_components(fin, bs, cf, bs_prev, log_list_calc)
        if all(map(np.isfinite, [c.get('ebitda'), c.get('revenue')])) and c['revenue'] > 1e-9:
            res = c['ebitda'] / c['revenue']
            log_list_calc.append(f"    Method 5 Component Calc (EBITDA Margin): {format_currency(c['ebitda'],'',0)} / {format_currency(c['revenue'],'',0)} = {format_percent(res,2)}")
            return res
        log_list_calc.append(f"    Method 5 Component Failed (Components: EBITDA={np.isfinite(c.get('ebitda'))}, Revenue={np.isfinite(c.get('revenue')) and c['revenue'] > 1e-9})")
        return np.nan

    # --- List of Methods to Try ---
    # Structure: [ (Display Name, Calculation Function, Needs Tax Rate?, Simple Formula String), ... ]
    methods_to_try = [
        ("EBITDA-TaxProv-WC-Capex Avg", calc_method0, False, "Avg(EBITDA - TaxProv - WC - Capex)"),
        ("NOPAT Avg", calc_method1, True, "Avg(EBIT*(1-T) + Depr - Capex - WC)"),
        ("CFO+IntAdj Avg", calc_method2, True, "Avg(CFO - Capex + IntExp*(1-T))"),
        ("CFO-CapEx Avg", calc_method3, False, "Avg(CFO - Capex)"),
        # Fallback methods based on averages of components
        ("Avg(EBITDA-TaxProv)", calc_method4, False, "Avg(EBITDA - Tax Provision)"),
        ("Avg(EBITDAMargin)*RevT0", calc_method5, False, "Avg(EBITDA/Revenue) * Revenue(T-0)"),
    ]

    # --- Iterate Through Methods and Calculate Averages ---
    for method_name, calc_func, needs_tax_rate, formula_str in methods_to_try:
        log_list_parent.append(f"  Attempting Basis Method: {method_name} (Target {years_to_average} yrs)...")
        yearly_values = []
        valid_count = 0
        log_details_yearly = [] # Store T-x: Value strings for summary

        # Special handling for margin method (needs T-0 Revenue later)
        is_margin_method = (method_name == "Avg(EBITDAMargin)*RevT0")
        rev_t0 = np.nan
        if is_margin_method:
             if 0 < financials_df.shape[1]: # Check if T-0 fin exists
                 rev_t0 = safe_get(financials_df.iloc[:, 0], ["Total Revenue", "Revenue", "Net Revenue"], "Income Stmt (T-0 for Margin Basis)", default=np.nan)
                 if not np.isfinite(rev_t0) or rev_t0 <= 1e-9:
                      log_list_parent.append(f"    ❌ Cannot use '{method_name}': Invalid T-0 Revenue ({format_currency(rev_t0,'',0)}). Skipping.")
                      continue # Skip this method if T-0 Rev is bad
                 else:
                      log_list_parent.append(f"    (Info for Margin Method: Using T-0 Revenue = {format_currency(rev_t0,'',0)})")
             else:
                 log_list_parent.append(f"    ❌ Cannot use '{method_name}': Missing T-0 financial data. Skipping.")
                 continue


        # Loop over the number of years available or target years, whichever is smaller
        for i in range(years_to_average):
             log_list_parent.append(f"  --- Calculating Components for T-{i} ({method_name}) ---")
             calc_log_this_year = [] # Capture logs from component fetching for this year/method

             # Check if enough data exists for this specific year 'i'
             if i >= financials_df.shape[1] or i >= cashflow_df.shape[1] or i >= balance_sheet_df.shape[1]:
                 log_list_parent.append(f"    Skipping T-{i}: Data unavailable in one or more statements.")
                 yearly_values.append(np.nan) # Add NaN if data missing for this year
                 log_details_yearly.append(f"T-{i}: N/A")
                 continue # Skip to next year

             fin = financials_df.iloc[:, i]; bs = balance_sheet_df.iloc[:, i]; cf = cashflow_df.iloc[:, i]
             # Safely get previous BS column
             bs_prev = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None

             # --- Call the appropriate calculation function ---
             val = np.nan # Default to NaN
             try:
                 # Pass the specific log list for this calculation
                 if needs_tax_rate:
                     val = calc_func(fin, bs, cf, bs_prev, tax_rt, calc_log_this_year)
                 else:
                     # Handle proxy methods needing fewer args (pass None for unused)
                     if method_name == "Avg(EBITDA-TaxProv)":
                         val = calc_func(fin, None, cf, None, calc_log_this_year)
                     elif method_name == "Avg(EBITDAMargin)*RevT0":
                          val = calc_func(fin, None, cf, None, calc_log_this_year) # This calculates the MARGIN here
                     else: # Standard methods not needing tax
                          val = calc_func(fin, bs, cf, bs_prev, calc_log_this_year)

             except Exception as e:
                 log_list_parent.append(f"    ❌ Error calculating T-{i} value for {method_name}: {e}")
                 log_list_parent.extend(calc_log_this_year) # Add any logs captured before error
                 val = np.nan # Ensure it's NaN on error

             log_list_parent.extend(calc_log_this_year) # Add logs from successful calculation
             yearly_values.append(val)
             # Format based on whether it's a margin or FCFF value
             formatted_val = format_percent(val, 1) if is_margin_method else format_currency(val, '', 0)
             log_details_yearly.append(f"T-{i}: {formatted_val}")
             if np.isfinite(val): valid_count += 1


        log_list_parent.append(f"  Finished {method_name} loop. Summary:")
        log_list_parent.append(f"    Values Calculated: [{', '.join(log_details_yearly)}]")

        if valid_count > 0:
            avg_val = np.nanmean(yearly_values) # nanmean ignores NaNs

            # If it's the margin method, calculate the final FCFF basis now
            final_basis_value = avg_val
            if is_margin_method:
                final_basis_value = avg_val * rev_t0 # Avg Margin * Rev T0
                log_list_parent.append(f"    Average Margin ({valid_count} valid yrs): {format_percent(avg_val, 2)}")
                log_list_parent.append(f"    FCFF Basis = Avg Margin * Rev T0 = {format_currency(final_basis_value, '', 0)}")
            else: # Standard FCFF averaging methods
                log_list_parent.append(f"    Average FCFF ({valid_count} valid yrs out of {years_to_average} attempted): {format_currency(avg_val, '', 0)}")


            # Check if the *final calculated basis value* is positive
            if np.isfinite(final_basis_value) and final_basis_value > 1e-9: # Check for strictly positive average/basis
                log_list_parent.append(f"    ✅ Resulting Average/Basis is positive.")
                # Store the successful result
                positive_bases[method_name] = {
                    'avg_fcff': final_basis_value,
                    'valid_years': valid_count,
                    'formula': formula_str
                }
            else:
                log_list_parent.append(f"    ❌ Resulting Average/Basis is non-positive or invalid. Cannot use this method.")
        else:
            log_list_parent.append(f"    ❌ No valid yearly values calculated for this method over the {years_to_average} years. Cannot use this method.")

    # Return the dictionary of successful methods
    return positive_bases

# --- End FCFF Helper Functions ---


# %% [markdown]
# ## 3. User Input & Dynamic Fetching: Ticker, Rf, Rm, Benchmarks, Beta, Spread

# %%
input_sources = {}

# --- Get Ticker Symbol ---
user_ticker = input(f"Enter stock ticker (e.g., MSFT, AAPL, RELIANCE.NS) [Default: {DEFAULT_TICKER}]: ")
if not user_ticker: user_ticker = DEFAULT_TICKER
ticker_symbol = user_ticker.strip().upper()
display(Markdown(f"**Analyzing Ticker:** `{ticker_symbol}`"))

# ----> DETECT IF INDIAN STOCK for currency formatting <----
is_indian_stock = ticker_symbol.endswith((".NS", ".BO")) # Include BSE
if is_indian_stock:
    print("Info: Detected Indian stock (.NS or .BO). Using ₹/Lakh/Crore formatting.")

# --- Determine RF Ticker based on Stock Ticker ---
rf_proxy_ticker = "^TNX" # Default US 10Y
if is_indian_stock: rf_proxy_ticker = "^NS10"; print(f"Info: Using Indian Rf proxy: {rf_proxy_ticker}")
elif ticker_symbol.endswith(".L"): rf_proxy_ticker = "^FTLC"; print(f"Info: Using UK Rf proxy: {rf_proxy_ticker}") # Example UK
# Add more regions as needed

# --- Get Risk-Free Rate (Rf) ---
print("\n--- CAPM Inputs ---")
dynamic_rf_rate = try_get_dynamic_rf(rf_proxy_ticker)
rf_prompt = f"Enter Risk-Free Rate (Rf) as percentage (e.g., '4.5' for 4.5%). Fetched {rf_proxy_ticker}?"
suggested_rf_val = dynamic_rf_rate * 100 if dynamic_rf_rate is not None else DEFAULT_RF_RATE_SUGGESTION
# Get input as percentage value (e.g., 4.5)
risk_free_rate_input_pct = get_validated_float_input(rf_prompt, default_value=f"{suggested_rf_val:.2f}", allow_blank=False, min_val=0.0, max_val=20.0, is_percentage=True)
# Convert to decimal for calculations
risk_free_rate_input = risk_free_rate_input_pct / 100.0
input_sources['Rf'] = 'Dynamic Fetch (User Confirmed/Adjusted)' if dynamic_rf_rate is not None else f'User Input (Default Suggestion: {DEFAULT_RF_RATE_SUGGESTION:.1f}%)'
display(Markdown(f"**Using Risk-Free Rate (Rf):** {format_percent(risk_free_rate_input)} (`Source: {input_sources['Rf']}`)"))

# --- Get Expected Market Return (Rm) ---
suggested_mrp = DEFAULT_MARKET_RISK_PREMIUM
suggested_rm = risk_free_rate_input + suggested_mrp
rm_prompt = f"Enter Expected Market Return (Rm) as percentage (e.g., '10'). Suggestion based on Rf + {format_percent(suggested_mrp)} MRP is {format_percent(suggested_rm)}"
while True:
    # Get input as percentage value (e.g., 10.0)
    market_return_input_pct = get_validated_float_input(rm_prompt, default_value=f"{suggested_rm * 100:.1f}", allow_blank=False, min_val=-10.0, max_val=35.0, is_percentage=True)
    # Convert to decimal for calculations
    market_return_input = market_return_input_pct / 100.0
    if market_return_input >= risk_free_rate_input: break
    else: print(f"❌ Error: Rm ({format_percent(market_return_input)}) must be >= Rf ({format_percent(risk_free_rate_input)}). Re-enter.")
market_risk_premium_input = market_return_input - risk_free_rate_input
input_sources['Rm'] = 'User Input'; input_sources['MRP'] = 'Derived'
display(Markdown(f"**Using Expected Market Return (Rm):** {format_percent(market_return_input)} (`Source: {input_sources['Rm']}`)"))
display(Markdown(f"**↳ Implied Market Risk Premium (MRP):** {format_percent(market_risk_premium_input)} (`Source: {input_sources['MRP']}`)"))

# --- Beta (Placeholder) ---
beta_input = None; input_sources['Beta'] = 'Pending'
# --- Credit Spread (Placeholder) ---
user_credit_spread = None; input_sources['Credit Spread'] = 'Pending'

# --- Get Industry Benchmark Ratios (with fallback defaults) ---
print("\n--- Industry Ratio Benchmarks ---")
display(Markdown("Enter benchmarks or leave blank for generic fallbacks. *(Find industry averages online)*"))
user_benchmarks = {}; benchmark_sources = {}
print("Provide benchmarks:")
for ratio_key in REQUIRED_BENCHMARK_RATIOS:
    default_val = DEFAULT_BENCHMARK_RATIOS.get(ratio_key, None)
    prompt = f"  Enter benchmark for {ratio_key}"
    is_percent_type = ratio_key in ['ROE', 'ROA', 'ROCE'] # These should be entered as percentages
    min_b, max_b = (-100, 200) if is_percent_type else (-100, 500) # Adjust bounds if needed
    # Use is_percentage flag for input prompt clarity
    bench_val_input = get_validated_float_input(prompt, default_value=f"{default_val:.2f}" if default_val is not None else None, allow_blank=True, min_val=min_b, max_val=max_b, is_percentage=is_percent_type)

    if bench_val_input is not None:
        # Store the value as entered (e.g., 15.0 for 15%)
        user_benchmarks[ratio_key] = bench_val_input; benchmark_sources[ratio_key] = 'User Input'
        # Format for display
        display_val = f"{bench_val_input:.1f}%" if is_percent_type else f"{bench_val_input:.2f}"
        # Special case for Asset Turnover (not a %)
        if ratio_key == 'Asset Turnover': display_val = f"{bench_val_input:.2f}"
        display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: User`)"))
    elif default_val is not None:
         # Store the default value (already in correct format, e.g., 19.05)
         user_benchmarks[ratio_key] = default_val; benchmark_sources[ratio_key] = 'Fallback Default'
         display_val = f"{default_val:.1f}%" if is_percent_type else f"{default_val:.2f}"
         if ratio_key == 'Asset Turnover': display_val = f"{default_val:.2f}"
         display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: Default`)"))
    else:
        user_benchmarks[ratio_key] = np.nan # Store NaN if no input and no default
        benchmark_sources[ratio_key] = 'Not Available'
        display(Markdown(f"  - **{ratio_key} Benchmark:** Not Provided / No Default"))
input_sources['Benchmarks'] = 'User Input or Default'


# %% [markdown]
# ## 4. Data Acquisition (Fetching Extended History)
# (No changes needed in this section's logic based on the request, keeping previous robust version)

# %%
stock_data = {}
try:
    stock = yf.Ticker(ticker_symbol)
    info = stock.info
    if not info or not info.get('regularMarketPrice'):
        print(f"Warning: Limited info for {ticker_symbol}. Trying history...")
        hist_check = stock.history(period="5d");
        if hist_check.empty: raise ValueError(f"Ticker '{ticker_symbol}' invalid/no data.")
        else: print(f"Warning: Proceeding with limited data."); info = {} if not isinstance(info, dict) else info
    stock_data['info'] = info

    # Fetch statements - use financials/balance_sheet/cashflow
    stock_data['financials'] = stock.financials # Annual usually
    stock_data['balance_sheet'] = stock.balance_sheet
    stock_data['cashflow'] = stock.cashflow

    # Sometimes quarterly data is fetched by default, try forcing annual if empty
    if stock_data['financials'].empty:
        print("Warning: Annual financials empty, trying quarterly...")
        stock_data['financials'] = stock.quarterly_financials
    if stock_data['balance_sheet'].empty:
        print("Warning: Annual balance sheet empty, trying quarterly...")
        stock_data['balance_sheet'] = stock.quarterly_balance_sheet
    if stock_data['cashflow'].empty:
        print("Warning: Annual cashflow empty, trying quarterly...")
        stock_data['cashflow'] = stock.quarterly_cashflow

    # Final Check
    missing_statements = []
    if stock_data['financials'].empty: missing_statements.append("Financials")
    if stock_data['balance_sheet'].empty: missing_statements.append("Balance Sheet")
    if stock_data['cashflow'].empty: missing_statements.append("Cash Flow")
    if missing_statements:
       print(f"⚠️ Warning: Could not retrieve essential statement(s): {', '.join(missing_statements)} for {ticker_symbol}. DCF unreliable/likely to fail.")

    stock_data['history'] = stock.history(period="2y") # For main price chart
    stock_data['history_long'] = stock.history(period="4y", interval="1d") # For stock CAGR calc
    if not stock_data['history_long'].empty:
        print(f"Info: Fetched long history (4y daily). Columns available: {list(stock_data['history_long'].columns)}")
    else:
         print(f"⚠️ Warning: Could not fetch long history (4y daily) for {ticker_symbol}. Stock CAGR calc may fail.")
    if stock_data['history_long'].empty or len(stock_data['history_long']) < 2:
        print(f"⚠️ Warning: Fetched long history has < 2 data points. Stock CAGR calc will fail.")

    company_name = info.get("longName", info.get("shortName", ticker_symbol))
    currency_code = info.get("currency", DEFAULT_CURRENCY_SYMBOL)
    if not isinstance(currency_code, str) or len(currency_code)>5 : currency_code=DEFAULT_CURRENCY_SYMBOL
    currency_symbol = "₹" if is_indian_stock else info.get("symbol", currency_code)

    # Robust price fetching
    current_price = safe_get(info, ['currentPrice', 'regularMarketPrice', 'previousClose', 'lastPrice', 'open', 'dayHigh', 'dayLow'], 'Info', default=np.nan)
    price_source = "yfinance Info"
    if not np.isfinite(current_price) and stock_data['history'] is not None and not stock_data['history'].empty:
        # Try last close from 2y history
        if 'Close' in stock_data['history'].columns and not stock_data['history']['Close'].empty:
            last_close_hist = stock_data['history']['Close'].iloc[-1]
            if pd.notna(last_close_hist) and last_close_hist > 0: current_price = last_close_hist; price_source = "yfinance History (Last Close)"
        # Fallback to last adj close
        elif 'Adj Close' in stock_data['history'].columns and not stock_data['history']['Adj Close'].empty:
             last_adj_close_hist = stock_data['history']['Adj Close'].iloc[-1]
             if pd.notna(last_adj_close_hist) and last_adj_close_hist > 0: current_price = last_adj_close_hist; price_source = "yfinance History (Last Adj Close)"

    if not np.isfinite(current_price): print(f"⚠️ Warning: Could not determine current price."); current_price = np.nan

    shares_outstanding = safe_get(info, ['sharesOutstanding'], 'Info', default=0)
    if not np.isfinite(shares_outstanding) or shares_outstanding <= 0: # Try getting from latest BS if info fails
        print("Warning: Shares outstanding from info is zero/invalid. Trying Balance Sheet...")
        if stock_data['balance_sheet'] is not None and not stock_data['balance_sheet'].empty:
            # Common keys for shares outstanding on BS
            shares_keys_bs = ['Share Issued', 'Ordinary Shares Number', 'Common Stock Equity'] # Add more if needed
            shares_bs = safe_get(stock_data['balance_sheet'].iloc[:,0], shares_keys_bs, 'Balance Sheet', default=0)
            if shares_bs is not None and np.isfinite(shares_bs) and shares_bs > 0:
                 shares_outstanding = shares_bs
                 print(f"Info: Using shares outstanding from Balance Sheet: {shares_outstanding:,.0f}")
            else:
                 print("Warning: Could not find valid shares outstanding on Balance Sheet either.")
                 shares_outstanding = 0 # Explicitly set to 0 if not found
        else: print("Warning: Balance sheet empty or invalid, cannot check for shares outstanding.")
        shares_outstanding = 0 if not np.isfinite(shares_outstanding) else shares_outstanding # Ensure it's 0 if still invalid

    beta_input = safe_get(info, ['beta', 'beta3Year'], 'Info', default=None) # Try 3-year beta too
    if beta_input is not None and np.isfinite(beta_input):
        input_sources['Beta'] = 'Dynamic Fetch'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))
    else:
         print(f"⚠️ WARNING: Beta not found dynamically or invalid.") # Changed to Warning
         beta_input_user = get_validated_float_input(f"Enter Beta for {ticker_symbol} (or leave blank to use 1.0):", default_value="1.0", allow_blank=True, min_val=-1.0, max_val=4.0)
         beta_input = float(beta_input_user) if beta_input_user is not None else 1.0
         info['beta'] = beta_input # Store it back for consistency
         input_sources['Beta'] = 'User Input' if beta_input_user is not None else 'Default (1.0)'
         display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))


    market_cap = safe_get(info, ['marketCap'], 'Info', default=0)
    mcap_source = "yfinance Info"
    # Recalculate Market Cap if possible and if fetched value is invalid
    if (not np.isfinite(market_cap) or market_cap <= 0) and shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Calculated (Price * Shares)"; print(f"Info: Market Cap (re)calculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}")
    elif not np.isfinite(market_cap) or market_cap <=0: mcap_source = "Unavailable"; print(f"⚠️ Warning: Market Cap unavailable."); market_cap = np.nan

    sector = info.get("sector", "N/A"); industry = info.get("industry", "N/A")
    summary = info.get("longBusinessSummary", "No summary available.")

    display(Markdown(f"**Company:** {company_name}")); display(Markdown(f"**Sector:** {sector} | **Industry:** {industry}"))
    display(Markdown(f"**Currency:** {currency_code} ({currency_symbol})"));
    display(Markdown(f"**Current Market Price:** {format_currency(current_price, currency_symbol, is_indian=is_indian_stock)} (`Source: {price_source}`)"))
    display(Markdown(f"**Market Cap:** {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)} (`Source: {mcap_source}`)"))
    display(Markdown(f"**Shares Outstanding:** {shares_outstanding:,.0f}"))


except Exception as e:
    print(f"❌ CRITICAL ERROR fetching data for {ticker_symbol}: {e}")
    print(traceback.format_exc()) # Print detailed traceback
    sys.exit("Cannot proceed due to data fetching error.")

# Essential Data Checks After Fetching
if not isinstance(shares_outstanding, (int, float)) or shares_outstanding <= 0:
    shares_outstanding_input = get_validated_float_input(f"CRITICAL: Shares Outstanding invalid ({shares_outstanding}). Enter value:", allow_blank=False, min_val=1)
    shares_outstanding = shares_outstanding_input
    if 'info' in stock_data and isinstance(stock_data['info'], dict): stock_data['info']['sharesOutstanding'] = shares_outstanding # Update info dict too
    print(f"Info: Using user-provided shares outstanding: {shares_outstanding:,.0f}")

if beta_input is None or not np.isfinite(beta_input):
    beta_input = 1.0 # Default to 1.0 if it somehow became invalid after input/fetch
    print("CRITICAL WARNING: Beta became invalid after fetch/input. Defaulting to 1.0. Review WACC.")
    input_sources['Beta'] = 'Default (1.0) - Invalid Fallback'

if not np.isfinite(market_cap):
    print(f"⚠️ WARNING: Market Cap unavailable/invalid. WACC weights will use Book Equity fallback or placeholder.")
    # Recalculate if possible, otherwise keep as NaN
    if shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Recalculated"; print(f"Info: Market Cap recalculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}")
    else:
        market_cap = np.nan # Ensure it's NaN if calculation failed

if not np.isfinite(current_price):
    print(f"⚠️ WARNING: Current Price unavailable/invalid. Valuation comparison and some ratios will be N/A.")


# %% [markdown]
# ## 5. WACC Calculation
# (No changes needed in this section's logic based on the request, keeping previous robust version)

# %%
# WACC Function Definition
def calculate_wacc(stock_info_dict, financials_df, balance_sheet_df,
                   user_rf, user_mrp, user_beta, input_sources_dict,
                   currency_sym, is_indian):
    global user_credit_spread # Allow modification of global variable
    display(Markdown("### Calculating WACC..."))
    components = {}
    calculation_log = []
    calculation_log.append(f"Input Sources: Rf='{input_sources_dict.get('Rf')}', MRP='{input_sources_dict.get('MRP')}', Beta='{input_sources_dict.get('Beta')}'")

    # Cost of Equity
    beta_val = user_beta
    if beta_val <= 0: calculation_log.append(f"⚠️ Warning: Non-positive Beta ({beta_val:.2f}) used in CAPM.")
    components['beta'] = beta_val; components['risk_free_rate'] = user_rf; components['market_risk_premium'] = user_mrp
    cost_of_equity = user_rf + beta_val * user_mrp
    # Basic sanity check on CoE (e.g., 0% to 50%)
    if not (0 < cost_of_equity < 0.50): calculation_log.append(f"⚠️ Warning: Calculated Cost of Equity ({format_percent(cost_of_equity)}) seems unusual.")
    components['cost_of_equity'] = cost_of_equity
    calculation_log.append(f"Inputs -> Rf: {format_percent(user_rf)}, Beta: {beta_val:.2f}, MRP: {format_percent(user_mrp)}")
    calculation_log.append(f"↳ Cost of Equity (Ke) [CAPM]: {format_percent(cost_of_equity)}")

    # Cost of Debt & Tax Rate
    cost_of_debt_pre_tax = np.nan; tax_rate = np.nan; after_tax_cost_of_debt = np.nan
    total_debt = 0.0; debt_calculation_method = "N/A"; tax_rate_source = "N/A"

    statements_available = (financials_df is not None and not financials_df.empty and
                            balance_sheet_df is not None and not balance_sheet_df.empty)

    if statements_available:
        latest_bs = balance_sheet_df.iloc[:, 0]; latest_fin = financials_df.iloc[:, 0]
        # --- Total Debt Calculation ---
        debt_keys_long = ['Total Debt Non Current', 'Long Term Debt', 'Long Term Debt And Capital Lease Obligation']
        debt_keys_short = ['Total Debt Current', 'Current Debt', 'Short Term Debt', 'Current Debt And Capital Lease Obligation', 'Short Long Term Debt']
        long_term_debt = safe_get(latest_bs, debt_keys_long , 'Balance Sheet (Long Debt)', default=0)
        short_term_debt = safe_get(latest_bs, debt_keys_short , 'Balance Sheet (Short Debt)', default=0)
        # Ensure debt components are numeric and finite, default to 0 if not
        long_term_debt = float(long_term_debt) if np.isfinite(long_term_debt) else 0.0
        short_term_debt = float(short_term_debt) if np.isfinite(short_term_debt) else 0.0
        total_debt = long_term_debt + short_term_debt
        components['debt_value_used'] = total_debt
        calculation_log.append(f"Latest BS Debt -> Long Term: {format_currency(long_term_debt, '', 0, is_indian)}, Short Term: {format_currency(short_term_debt, '', 0, is_indian)} -> Total Debt: {format_currency(total_debt, '', 0, is_indian)}")


        # --- Pre-Tax Cost of Debt (Kd) ---
        interest_expense_keys = ['Interest Expense', 'Interest Expense Net']
        interest_expense = abs(safe_get(latest_fin, interest_expense_keys, 'Income Statement (Interest)', default=0))
        interest_expense = float(interest_expense) if np.isfinite(interest_expense) else 0.0
        calculation_log.append(f"Latest Income Stmt -> Interest Expense: {format_currency(interest_expense, '', 0, is_indian)}")


        implied_rate_calculated = False
        if total_debt > 1000 and interest_expense > 0: # Require minimum debt and positive interest
             implied_rate = interest_expense / total_debt; implied_rate_calculated = True
             calculation_log.append(f"Implied Kd = Interest / Debt = {format_currency(interest_expense,'',0,is_indian)} / {format_currency(total_debt,'',0,is_indian)} = {format_percent(implied_rate)}")
             if MIN_COST_OF_DEBT <= implied_rate <= MAX_COST_OF_DEBT:
                  cost_of_debt_pre_tax = implied_rate; debt_calculation_method = "Implied (Interest/Debt)";
                  calculation_log.append(f"↳ Using Implied Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)} (Within Bounds)")
             else: calculation_log.append(f"⚠️ Implied Kd ({format_percent(implied_rate)}) out of bounds ({format_percent(MIN_COST_OF_DEBT)}-{format_percent(MAX_COST_OF_DEBT)}). Falling back."); implied_rate_calculated=False # Reset flag
        elif total_debt > 1000: calculation_log.append(f"ℹ️ Found Debt but zero/invalid Interest Exp. Cannot use implied Kd. Falling back.")
        else: calculation_log.append(f"ℹ️ Negligible/Zero Debt. Cannot use implied Kd.")


        # Fallback Kd: Rf + Spread
        if not implied_rate_calculated:
             if total_debt > 1000: # Only apply default if debt exists and implied failed
                 debt_calculation_method = "Default (Rf+Spread)"
                 if user_credit_spread is None:
                     spread_prompt = f"Implied Kd failed/NA for Debt={format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd"
                     # Get spread as percentage points (e.g., 1.5)
                     user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=15.0, is_percentage=True)
                     user_credit_spread = user_credit_spread_pct / 100.0 # Convert to decimal
                     input_sources_dict['Credit Spread'] = 'User Input (for Default Kd)'
                 else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
                 cost_of_debt_pre_tax = user_rf + user_credit_spread
                 original_default_kd = cost_of_debt_pre_tax
                 # Bound the default cost of debt as well
                 cost_of_debt_pre_tax = min(max(cost_of_debt_pre_tax, MIN_COST_OF_DEBT), MAX_COST_OF_DEBT)
                 calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(original_default_kd)}")
                 if abs(cost_of_debt_pre_tax - original_default_kd) > 1e-6 : calculation_log.append(f"↳ Bounded Default Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)}")
                 else: calculation_log.append(f"↳ Using Default Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)}")
                 calculation_log.append(f"   (Spread Source: {input_sources_dict['Credit Spread']})")

             else: # Negligible/Zero Debt
                 debt_calculation_method = "Zero Debt"; cost_of_debt_pre_tax = 0.0; input_sources_dict['Credit Spread'] = 'N/A (Zero Debt)';
                 calculation_log.append(f"ℹ️ Negligible/Zero Debt ({format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}). Setting Kd = 0%.")

        # --- Tax Rate Calculation ---
        tax_prov_keys = ['Tax Provision', 'Income Tax Expense']
        pretax_inc_keys = ['Pretax Income', 'EBT', 'Earnings Before Tax']
        income_tax_expense = safe_get(latest_fin, tax_prov_keys, 'Income Statement (Tax Exp)', default=0)
        pretax_income = safe_get(latest_fin, pretax_inc_keys, 'Income Statement (Pretax Inc)', default=0)
        # Ensure numeric/finite, default 0 if not
        income_tax_expense = float(income_tax_expense) if np.isfinite(income_tax_expense) else 0.0
        pretax_income = float(pretax_income) if np.isfinite(pretax_income) else 0.0
        calculation_log.append(f"Latest Income Stmt -> Tax Expense: {format_currency(income_tax_expense, '', 0, is_indian)}, Pretax Income: {format_currency(pretax_income, '', 0, is_indian)}")


        # Attempt to calculate EBT from EBIT if EBT is zero/negative but tax exists
        if pretax_income <= 1 and income_tax_expense > 0: # Use 1 instead of 0 to handle rounding
             ebit_val = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=0);
             ebit_val = float(ebit_val) if np.isfinite(ebit_val) else 0.0
             # Use interest expense already fetched
             pretax_income_from_ebit = ebit_val - interest_expense
             if pretax_income_from_ebit > 1: # Check if derived EBT is positive
                 pretax_income = pretax_income_from_ebit
                 calculation_log.append(f"ℹ️ Pretax Income was near zero/negative. Used EBIT-Interest as proxy: {format_currency(pretax_income, symbol=currency_sym, is_indian=is_indian)}")


        if pretax_income > 1 and income_tax_expense >= 0: # Require positive pretax income
            calculated_tax_rate = income_tax_expense / pretax_income;
            # Bound the calculated tax rate
            tax_rate = min(max(calculated_tax_rate, MIN_TAX_RATE), MAX_TAX_RATE); tax_rate_source = "Derived & Bounded"
            if abs(tax_rate - calculated_tax_rate) > 0.001:
                 tax_rate_source = f"Derived ({format_percent(calculated_tax_rate)}), Bounded to {format_percent(MIN_TAX_RATE)}-{format_percent(MAX_TAX_RATE)}"
                 calculation_log.append(f"Calculated Tax Rate = Tax Exp / Pretax Inc = {format_percent(calculated_tax_rate)}")
                 calculation_log.append(f"↳ Effective Tax Rate: {format_percent(tax_rate)} ({tax_rate_source})")
            else:
                 calculation_log.append(f"Calculated Tax Rate = Tax Exp / Pretax Inc = {format_percent(calculated_tax_rate)}")
                 calculation_log.append(f"↳ Effective Tax Rate: {format_percent(tax_rate)} ({tax_rate_source})")
        else:
             tax_rate = DEFAULT_TAX_RATE; tax_rate_source = "Default (Derived Failed)";
             calculation_log.append(f"⚠️ Cannot derive tax rate (Pretax Inc <= 0 or Tax Exp < 0). Using default: {format_percent(tax_rate)}")

    else: # Financial statements empty/invalid
         calculation_log.append("⚠️ Financial statements empty/invalid. Using defaults for Kd and Tax Rate."); debt_calculation_method = "Default (No Fin.)"; tax_rate_source = "Default (No Fin.)"
         if user_credit_spread is None:
              spread_prompt = f"No financials. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd"
              user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=15.0, is_percentage=True)
              user_credit_spread = user_credit_spread_pct / 100.0; input_sources_dict['Credit Spread'] = 'User Input (No Fin.)'
         else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
         cost_of_debt_pre_tax = user_rf + user_credit_spread;
         original_default_kd = cost_of_debt_pre_tax
         cost_of_debt_pre_tax = min(max(cost_of_debt_pre_tax, MIN_COST_OF_DEBT), MAX_COST_OF_DEBT) # Bound default
         tax_rate = DEFAULT_TAX_RATE; total_debt = 0; components['debt_value_used'] = 0
         calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(original_default_kd)}")
         if abs(cost_of_debt_pre_tax - original_default_kd) > 1e-6 : calculation_log.append(f"↳ Bounded Default Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)}")
         else: calculation_log.append(f"↳ Using Default Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)}")
         calculation_log.append(f"Using Default Tax Rate: {format_percent(tax_rate)}")

    # Ensure components are valid numbers before calculating after-tax Kd
    components['cost_of_debt'] = cost_of_debt_pre_tax if np.isfinite(cost_of_debt_pre_tax) else 0.0
    components['debt_calc_method'] = debt_calculation_method
    components['tax_rate'] = tax_rate if np.isfinite(tax_rate) else DEFAULT_TAX_RATE
    components['tax_rate_source'] = tax_rate_source

    if np.isfinite(components['cost_of_debt']) and np.isfinite(components['tax_rate']):
        after_tax_cost_of_debt = components['cost_of_debt'] * (1 - components['tax_rate'])
    else:
        calculation_log.append(f"⚠️ Invalid Kd ({components['cost_of_debt']}) or Tax Rate ({components['tax_rate']}). Cannot calculate after-tax Kd.")
        after_tax_cost_of_debt = np.nan # Set to NaN if components invalid

    components['after_tax_cost_of_debt'] = after_tax_cost_of_debt
    if np.isfinite(after_tax_cost_of_debt) and components['cost_of_debt'] > 0 : calculation_log.append(f"↳ Cost of Debt (After-tax): {format_percent(after_tax_cost_of_debt)}")

    # Equity and Debt Weights
    market_cap_val = safe_get(stock_info_dict, ['marketCap'], 'Info', default=np.nan); equity_source = "Market Cap" if np.isfinite(market_cap_val) and market_cap_val > 0 else "N/A"
    equity_value_for_wacc = np.nan

    if np.isfinite(market_cap_val) and market_cap_val > 0:
        equity_value_for_wacc = market_cap_val
        equity_source = "Market Cap"
    else: # Market cap failed, try Book Equity
       calculation_log.append(f"⚠️ Market Cap unavailable/zero ({format_currency(market_cap_val, symbol=currency_sym, is_indian=is_indian)}). Falling back to Book Equity."); equity_source = "Book Equity Fallback"
       if balance_sheet_df is not None and not balance_sheet_df.empty:
           equity_keys = ['Total Stockholder Equity', 'Stockholders Equity', 'Total Equity Gross Minority Interest', 'Common Stock Equity']
           book_equity_val = safe_get(balance_sheet_df.iloc[:, 0], equity_keys, 'Balance Sheet (Book Equity)', default=np.nan)
           if np.isfinite(book_equity_val) and book_equity_val > 1.0: # Require positive book equity
              equity_value_for_wacc = book_equity_val;
              calculation_log.append(f"Using Book Equity ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}) as proxy for Market Equity in WACC weights.")
           else:
                equity_source = "Placeholder (Book Invalid)";
                calculation_log.append(f"⚠️ Book Equity also invalid/unavailable ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}).")
       else: equity_source = "Placeholder (No BS)"
       # Final placeholder if both Market and Book fail
       if not np.isfinite(equity_value_for_wacc):
            equity_value_for_wacc = 1.0; # Use 1.0 to avoid division by zero, but weights will be wrong
            equity_source = "Placeholder (Market/Book Failed)"
            calculation_log.append("⚠️ Market Cap & Book Equity unavailable/invalid. Using placeholder Equity Value (1.0). WACC Weights will be inaccurate.")

    components['equity_value_used'] = equity_value_for_wacc; components['equity_source'] = equity_source
    equity_weight = np.nan; debt_weight = np.nan; weights_source = "N/A";
    # Use the debt value calculated earlier
    total_capital = equity_value_for_wacc + total_debt
    if np.isfinite(total_capital) and total_capital > 1.0: # Require positive total capital
        equity_weight = equity_value_for_wacc / total_capital; debt_weight = total_debt / total_capital; weights_source = "Calculated"
        total_weight = equity_weight + debt_weight
        # Normalize weights if they don't sum to 1 (e.g., due to floating point)
        if abs(total_weight - 1.0) > 1e-6:
             calculation_log.append(f"ℹ️ Normalizing E/D weights (Sum was {total_weight:.4f})")
             equity_weight /= total_weight; debt_weight /= total_weight
    else: calculation_log.append(f"❌ ERROR: Total Capital ({format_currency(total_capital, symbol=currency_sym, is_indian=is_indian)}) invalid/zero. Using default weights (70/30)."); equity_weight = 0.7; debt_weight = 0.3; weights_source = "Default (Invalid Cap)"

    components['equity_weight'] = equity_weight; components['debt_weight'] = debt_weight; components['weights_source'] = weights_source
    calculation_log.append(f"Equity Value Used ({components['equity_source']}): {format_currency(components['equity_value_used'], symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"Debt Value Used ({components['debt_calc_method']}): {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"↳ Weights E/V: {format_percent(equity_weight)}, D/V: {format_percent(debt_weight)} (Source: {weights_source})")

    # Final WACC Calculation
    wacc = np.nan
    # Check all necessary components are finite numbers
    if all(map(np.isfinite, [equity_weight, cost_of_equity, debt_weight, after_tax_cost_of_debt])):
        wacc = (equity_weight * cost_of_equity) + (debt_weight * after_tax_cost_of_debt); original_wacc = wacc;
        # Apply bounds to WACC
        wacc = min(max(wacc, MIN_WACC), MAX_WACC)
        bound_note = ""
        if abs(wacc - original_wacc) > 0.0001 : bound_note = f"(Bounded from {format_percent(original_wacc)} to {format_percent(MIN_WACC)}-{format_percent(MAX_WACC)})"; calculation_log.append(f"⚠️ WACC ({format_percent(original_wacc)}) bounded.")
        calculation_log.append(f"▶️ Calculated WACC: {format_percent(wacc)} {bound_note}")
    else:
        calculation_log.append(f"❌ ERROR: Cannot calculate final WACC due to invalid components:")
        if not np.isfinite(equity_weight): calculation_log.append(f"   - Equity Weight: {equity_weight}")
        if not np.isfinite(cost_of_equity): calculation_log.append(f"   - Cost of Equity: {cost_of_equity}")
        if not np.isfinite(debt_weight): calculation_log.append(f"   - Debt Weight: {debt_weight}")
        if not np.isfinite(after_tax_cost_of_debt): calculation_log.append(f"   - After-Tax Cost of Debt: {after_tax_cost_of_debt}")
        wacc = np.nan

    components['final_wacc'] = wacc
    display(Markdown("```\n" + "\n".join(calculation_log) + "\n```"))
    # Plot WACC Donut Chart
    if np.isfinite(wacc) and wacc > 0 and np.isfinite(equity_weight) and np.isfinite(debt_weight):
        equity_contrib = equity_weight * cost_of_equity; debt_contrib = debt_weight * after_tax_cost_of_debt
        # Check if contributions are valid and sum approximately to WACC
        if np.isfinite(equity_contrib) and np.isfinite(debt_contrib) and abs(equity_contrib + debt_contrib - wacc) < 0.01:
           labels = ['Cost of Equity Contribution', 'Cost of Debt Contribution']; values = [equity_contrib, debt_contrib]
           fig_wacc_donut = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4, marker_colors=px.colors.qualitative.Pastel, hovertemplate='<b>%{label}</b><br>Contribution: %{value:.2%}<extra></extra>', textinfo='percent', textfont_size=12, sort=False)])
           fig_wacc_donut.update_layout(title_text=f'WACC Composition (Total: {format_percent(wacc)})', title_x=0.5, showlegend=True, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), margin=dict(l=20, r=20, t=60, b=50), height=350)
           fig_wacc_donut.show()
        else: display(Markdown(f"*(WACC component contributions invalid/don't sum ({format_percent(equity_contrib)} + {format_percent(debt_contrib)} != {format_percent(wacc)}). Cannot generate composition chart.)*"))
    elif not np.isfinite(wacc): display(Markdown(f"*(WACC calculation failed. Cannot generate composition chart)*"))
    else: display(Markdown(f"*(WACC zero/negative or weights invalid. Cannot generate composition chart)*"))

    return wacc, components

# Execute WACC Calculation
discount_rate, wacc_components = calculate_wacc(
    stock_data['info'], stock_data['financials'], stock_data['balance_sheet'],
    risk_free_rate_input, market_risk_premium_input, beta_input, input_sources,
    currency_symbol, is_indian_stock
)


# %% [markdown]
# ## 6. FCFF Calculation & Projection (V21.2 - User Choice for Basis)

# %%
# FCFF Function Definition (V21.2 - User Choice Logic Added)
def calculate_and_project_fcff_v21_user_choice(financials_df, balance_sheet_df, cashflow_df, history_long_df, # History needed for stock CAGR
                                   wacc_comps, disc_rate, forecast_yrs,
                                   final_user_rf, default_init_growth, # This is 6.5% default
                                   currency_sym, is_indian):
    display(Markdown("### Calculating & Projecting FCFF..."))
    projection_log = [] # Master log for this section
    projection_results = {
        'historical_fcffs_individual': {}, # Store individual yearly best estimates for CAGR calc
        'historical_revenues': {}, # Store recent historical revenues
        'current_fcff_basis': np.nan, # This will be the positive average basis chosen by user or fallback
        'projected_fcffs': [], 'growth_rates': [],
        'pv_fcffs': [], 'cumulative_pv_fcff': np.nan, 'terminal_growth_rate': np.nan,
        'terminal_value': np.nan, 'pv_terminal_value': np.nan, 'error': None,
        'initial_growth_rate': np.nan, 'growth_source': 'N/A',
        'fcff_basis_method': 'N/A' # Store method used for the POSITIVE average basis
    }

    # --- Pre-checks ---
    statements_valid = True
    if financials_df is None or financials_df.empty:
        projection_results['error'] = "Missing Financials Statement."; statements_valid = False
    if balance_sheet_df is None or balance_sheet_df.empty:
        projection_results['error'] = (projection_results['error'] + " Missing Balance Sheet.") if projection_results['error'] else "Missing Balance Sheet."
        statements_valid = False
    if cashflow_df is None or cashflow_df.empty:
        projection_results['error'] = (projection_results['error'] + " Missing Cash Flow.") if projection_results['error'] else "Missing Cash Flow."
        statements_valid = False

    if not statements_valid:
         display(Markdown(f"❌ ERROR: {projection_results['error']} Cannot proceed."))
         return projection_results # Exit early

    if not np.isfinite(disc_rate):
         projection_results['error'] = "Invalid WACC provided."; display(Markdown(f"❌ ERROR: {projection_results['error']}")); return projection_results # Exit early

    # --- Get Effective Tax Rate ---
    effective_tax_rate = DEFAULT_TAX_RATE; tax_rate_log_src = "Default"
    if isinstance(wacc_comps, dict) and 'tax_rate' in wacc_comps and np.isfinite(wacc_comps['tax_rate']):
        effective_tax_rate = wacc_comps['tax_rate']
        tax_rate_log_src = wacc_comps.get('tax_rate_source', '?')
    projection_log.append(f"Using Tax Rate from WACC: {format_percent(effective_tax_rate)} ({tax_rate_log_src})")

    # ==============================================================
    # PHASE 1: Calculate Historical Data (Individual Years for CAGR)
    # ==============================================================
    projection_log.append("\n--- Phase 1: Calculating Historical FCFF & Revenue (for CAGR Estimation) ---")
    projection_log.append("    (Uses best estimate for each individual year: NOPAT -> CFO+IntAdj -> CFO-Capex)")

    num_years_cagr = 3 # CAGR uses 3 years (T0 vs T3)
    max_hist_years_needed = num_years_cagr + 1 # Need 4 data points (T-0, T-1, T-2, T-3)
    num_statement_years_avail = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    historical_fcffs_list_individual = [] # Best estimate for each year T-0..T-3
    historical_revenues_list = [] # Newest first (index 0 = T-0)

    years_to_calc_hist = min(num_statement_years_avail, max_hist_years_needed)
    if years_to_calc_hist < max_hist_years_needed:
         projection_log.append(f"  ⚠️ Warning: Only {years_to_calc_hist} years of statement data available (needed {max_hist_years_needed} for full 3yr CAGR calculations). CAGR may be unreliable/fail.")

    for i in range(years_to_calc_hist):
        projection_log.append(f"\n--- Processing Historical Year T-{i} ---")
        # Check if data exists for this specific year 'i'
        if i >= financials_df.shape[1] or i >= balance_sheet_df.shape[1] or i >= cashflow_df.shape[1]:
            projection_log.append(f"  --> Historical FCFF T-{i} (for CAGR): N/A (Data Missing)")
            projection_log.append(f"  --> Historical Revenue T-{i} (for CAGR): N/A (Data Missing)")
            historical_fcffs_list_individual.append(np.nan)
            historical_revenues_list.append(np.nan)
            continue

        fin_col = financials_df.iloc[:, i]; bs_col = balance_sheet_df.iloc[:, i]; cf_col = cashflow_df.iloc[:, i]
        bs_prev_col = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None

        # Use the single-year fallback helper for the historical list needed for CAGR
        fcff_val, fcff_method_yr = _calculate_fcff_single_year_best_estimate(fin_col, bs_col, cf_col, bs_prev_col, effective_tax_rate, projection_log)
        historical_fcffs_list_individual.append(fcff_val)
        projection_log.append(f"  --> Historical FCFF T-{i} (for CAGR): {format_currency(fcff_val, '', 0)} (Method Used: {fcff_method_yr})")

        rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
        revenue_val = safe_get(fin_col, rev_keys, default=np.nan)
        historical_revenues_list.append(revenue_val)
        # Log revenue too for clarity
        projection_log.append(f"  --> Historical Revenue T-{i} (for CAGR): {format_currency(revenue_val, '', 0)}")


    # Pad with NaNs if data is shorter than needed for CAGR calculations later
    while len(historical_fcffs_list_individual) < max_hist_years_needed: historical_fcffs_list_individual.append(np.nan)
    while len(historical_revenues_list) < max_hist_years_needed: historical_revenues_list.append(np.nan)

    projection_results['historical_fcffs_individual'] = {f"T-{i}": v for i, v in enumerate(historical_fcffs_list_individual)}
    projection_results['historical_revenues'] = {f"T-{i}": v for i, v in enumerate(historical_revenues_list)}

    # ====================================================================
    # PHASE 2: Determine FCFF Starting Basis for Projection (User Choice)
    # ====================================================================
    projection_log.append("\n--- Phase 2: Determining FCFF Basis for Projection Start ---")
    fcff_start_basis = np.nan
    fcff_basis_method = "N/A"

    # --- Step 2a: Calculate all potential positive average bases ---
    # This function now returns a dict of successful methods only
    potential_bases = _calculate_all_fcff_bases_v21_4yr(
        financials_df, balance_sheet_df, cashflow_df, effective_tax_rate, projection_log
    )

    # --- Step 2b: Present options to user or use fallback ---
    chosen_method_details = None
    if potential_bases:
        projection_log.append("\n--- User Selection: Choose FCFF Projection Basis ---")
        display(Markdown("--- User Selection: Choose FCFF Projection Basis ---"))
        print("The following methods resulted in a positive average FCFF over the last ~4 years:")
        options_list = list(potential_bases.items()) # Convert dict items to list for indexed access
        for idx, (method_name, details) in enumerate(options_list):
            print(f"  {idx+1}. {method_name}: Avg FCFF = {format_currency(details['avg_fcff'], currency_sym, 2, is_indian)} ({details['valid_years']} valid yrs)")
            print(f"     Formula: {details['formula']}")

        while chosen_method_details is None:
             try:
                 choice = input(f"Enter the number of the method you want to use as the projection starting basis [1-{len(options_list)}]: ")
                 choice_idx = int(choice) - 1
                 if 0 <= choice_idx < len(options_list):
                     chosen_method_key, chosen_method_details = options_list[choice_idx]
                     fcff_start_basis = chosen_method_details['avg_fcff']
                     fcff_basis_method = f"{chosen_method_key} ({chosen_method_details['valid_years']}yr Avg - User Selected)"
                     projection_log.append(f"  ✅ User selected method {choice}: {chosen_method_key}")
                     projection_log.append(f"  Using projection start basis: {format_currency(fcff_start_basis, '', 0)}")
                     display(Markdown(f"**Using Basis:** {fcff_basis_method} = **{format_currency(fcff_start_basis, currency_sym, 2, is_indian)}**"))

                 else: print(f"❌ Invalid choice. Please enter a number between 1 and {len(options_list)}.")
             except ValueError: print("❌ Invalid input. Please enter a number.")

    else: # No averaging methods worked, try T-0 fallback
        projection_log.append("\n--- No Positive Average FCFF Found from Methods. Trying T-0 Fallback ---")
        # Recalculate T-0 individual best estimate (might already be in historical list, but recalculate for clarity here)
        if 0 < num_statement_years_avail: # Check if T-0 data exists
            fin_t0 = financials_df.iloc[:, 0]; bs_t0 = balance_sheet_df.iloc[:, 0]; cf_t0 = cashflow_df.iloc[:, 0]
            bs_t1 = balance_sheet_df.iloc[:, 1] if balance_sheet_df.shape[1] > 1 else None
            projection_log.append("\n--- Calculating T-0 Individual Estimate for Fallback Basis ---")
            fcff_t0_ind, method_t0_ind = _calculate_fcff_single_year_best_estimate(fin_t0, bs_t0, cf_t0, bs_t1, effective_tax_rate, projection_log)

            if np.isfinite(fcff_t0_ind) and fcff_t0_ind > 1e-9:
                 fcff_start_basis = fcff_t0_ind
                 fcff_basis_method = f"{method_t0_ind} (T-0 Only / Fallback)"
                 projection_log.append(f"  ✅ Fallback T-0 basis is positive: {format_currency(fcff_start_basis, '', 0)}")
                 projection_log.append(f"  Using projection start basis: {format_currency(fcff_start_basis, '', 0)}")
                 display(Markdown(f"**Using Fallback Basis:** {fcff_basis_method} = **{format_currency(fcff_start_basis, currency_sym, 2, is_indian)}**"))

            else:
                 projection_log.append(f"  ❌ Fallback T-0 Individual Estimate is non-positive/invalid: {format_currency(fcff_t0_ind)} (Method: {method_t0_ind}).")
                 fcff_basis_method = "No Positive Basis Found"
        else:
            projection_log.append("  ❌ Cannot attempt T-0 fallback basis: Missing T-0 statement data.")
            fcff_basis_method = "No Positive Basis Found (Missing T-0)"


    projection_results['current_fcff_basis'] = fcff_start_basis
    projection_results['fcff_basis_method'] = fcff_basis_method

    # --- Step 2c: Halt if no positive basis found after all attempts ---
    if not np.isfinite(fcff_start_basis):
        error_msg = f"Could not determine a valid POSITIVE FCFF basis for projection start (Final status: {fcff_basis_method}). Cannot proceed with projection."
        projection_results['error'] = error_msg
        projection_log.append(f"\n❌ HALTING: {error_msg}")
        # Display the log up to this point
        display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
        # Stop before growth calculation / projection
        return projection_results

    # ================================================================
    # PHASE 3: Estimate Growth Rates (Using Historical Data from Ph 1)
    # ================================================================
    projection_log.append("\n--- Phase 3: Estimating Initial Growth Rate (CAGR Geometric Mean) ---")
    # Uses the INDIVIDUAL historical FCFFs (T-0 vs T-3) calculated in Phase 1
    initial_growth_rate = np.nan # Start with NaN
    growth_source = "N/A"
    calculated_cagrs = {} # Store individual CAGRs {Name: value}
    valid_cagr_list = [] # Store only the valid CAGRs for GM calculation

    # 1. Calculate 3-Year Revenue CAGR (Conditional)
    rev_cagr = np.nan
    projection_log.append(f"\nCalculating Revenue {num_years_cagr}yr CAGR (T-0 vs T-{num_years_cagr})...")
    if len(historical_revenues_list) >= max_hist_years_needed:
        rev_end = historical_revenues_list[0]; rev_start = historical_revenues_list[num_years_cagr] # Index 3 is T-3
        projection_log.append(f"  Data Points -> Start (T-{num_years_cagr}): {format_currency(rev_start, '', 0, is_indian)}, End (T-0): {format_currency(rev_end, '', 0, is_indian)}")
        if pd.notna(rev_end) and math.isfinite(rev_end) and pd.notna(rev_start) and math.isfinite(rev_start) and rev_start > 1e-9:
            rev_cagr = calculate_cagr_modified(rev_start, rev_end, num_years_cagr)
            if pd.notna(rev_cagr) and math.isfinite(rev_cagr):
                 calculated_cagrs['Revenue'] = rev_cagr; valid_cagr_list.append(rev_cagr)
                 projection_log.append(f"  ✅ Revenue CAGR: {format_percent(rev_cagr)}")
            else: projection_log.append(f"  ❌ Revenue CAGR Calculation Failed (Result Invalid: {rev_cagr}).")
        elif not (pd.notna(rev_start) and math.isfinite(rev_start) and rev_start > 1e-9):
             projection_log.append(f"  ❌ Revenue CAGR Skipped (Invalid/Non-positive Start Data: {format_currency(rev_start, '', 0, is_indian)}).")
        else: # Start is valid, but end is not
             projection_log.append(f"  ❌ Revenue CAGR Skipped (Invalid End Data: {format_currency(rev_end, '', 0, is_indian)}).")
    else: projection_log.append(f"  ❌ Revenue CAGR Skipped (Insufficient data points: {len(historical_revenues_list)} < {max_hist_years_needed})")

    # 2. Calculate 3-Year FCFF CAGR (Conditional - uses INDIVIDUAL historical FCFFs)
    fcff_cagr = np.nan
    projection_log.append(f"\nCalculating FCFF {num_years_cagr}yr CAGR (T-0 vs T-{num_years_cagr})...")
    if len(historical_fcffs_list_individual) >= max_hist_years_needed:
        fcff_end = historical_fcffs_list_individual[0]; fcff_start = historical_fcffs_list_individual[num_years_cagr] # Index 3 is T-3
        projection_log.append(f"  Data Points -> Start (T-{num_years_cagr}): {format_currency(fcff_start, '', 0, is_indian)}, End (T-0): {format_currency(fcff_end, '', 0, is_indian)}")
        if pd.notna(fcff_end) and math.isfinite(fcff_end) and pd.notna(fcff_start) and math.isfinite(fcff_start) and fcff_start > 1e-9:
            fcff_cagr = calculate_cagr_modified(fcff_start, fcff_end, num_years_cagr)
            if pd.notna(fcff_cagr) and math.isfinite(fcff_cagr):
                 calculated_cagrs['FCFF'] = fcff_cagr; valid_cagr_list.append(fcff_cagr)
                 projection_log.append(f"  ✅ FCFF CAGR:    {format_percent(fcff_cagr)}")
            else: projection_log.append(f"  ❌ FCFF CAGR Calculation Failed (Result Invalid: {fcff_cagr}).")
        elif not (pd.notna(fcff_start) and math.isfinite(fcff_start) and fcff_start > 1e-9):
            projection_log.append(f"  ❌ FCFF CAGR Skipped (Invalid/Non-positive Start Data: {format_currency(fcff_start, '', 0, is_indian)}).")
        else: # Start is valid, but end is not
            projection_log.append(f"  ❌ FCFF CAGR Skipped (Invalid End Data: {format_currency(fcff_end, '', 0, is_indian)}).")
    else: projection_log.append(f"  ❌ FCFF CAGR Skipped (Insufficient data points: {len(historical_fcffs_list_individual)} < {max_hist_years_needed})")

    # 3. Calculate 3-Year Stock Price CAGR (Conditional - Robust Column V21)
    stock_cagr = np.nan; price_col = None
    projection_log.append(f"\nCalculating Stock Price {num_years_cagr}yr CAGR...")
    if history_long_df is not None and not history_long_df.empty and len(history_long_df) >= 2:
         # Prefer Adj Close, fallback to Close
         if 'Adj Close' in history_long_df.columns and history_long_df['Adj Close'].notna().any(): price_col = 'Adj Close'
         elif 'Close' in history_long_df.columns and history_long_df['Close'].notna().any(): price_col = 'Close'; projection_log.append("  ⚠️ Using 'Close' price for Stock CAGR ('Adj Close' not found or all NaN).")
         else: projection_log.append(f"  ❌ Stock CAGR Skipped (Neither 'Adj Close' nor 'Close' columns found or contain valid data)")

         if price_col and isinstance(history_long_df.index, pd.DatetimeIndex):
             projection_log.append(f"  Stock History Info: Using '{price_col}' column. Found {len(history_long_df)} rows, Date Range: {history_long_df.index.min().date()} to {history_long_df.index.max().date()}")
             try:
                 # Ensure data is sorted by date
                 history_long_df = history_long_df.sort_index()
                 # Find the actual end date and price (last valid non-NaN)
                 end_series = history_long_df[price_col].dropna()
                 if end_series.empty: raise ValueError("Price column has no valid data.")
                 end_date = end_series.index[-1]; end_price = end_series.iloc[-1]

                 # Find the target start date
                 start_date_target = end_date - pd.DateOffset(years=num_years_cagr)

                 # Find the closest available date >= target start date
                 start_series = history_long_df.loc[start_date_target : end_date - pd.Timedelta(days=1), price_col].dropna() # Look between target start and day before end
                 if start_series.empty:
                     # If nothing found after target, try looking slightly before as fallback
                     start_series = history_long_df.loc[:start_date_target, price_col].dropna()
                     if start_series.empty: raise ValueError(f"No valid price data found near target start date {start_date_target.date()}")
                     start_actual_date = start_series.index[-1]; start_price = start_series.iloc[-1]
                     projection_log.append(f"  ⚠️ Could not find data at or after target start date. Using closest prior date: {start_actual_date.date()}")
                 else:
                     start_actual_date = start_series.index[0]; start_price = start_series.iloc[0]


                 actual_years = (end_date - start_actual_date).days / 365.25
                 projection_log.append(f"  Stock CAGR Data Points: Start Date={start_actual_date.date()}, Start Price={start_price:.2f}, End Date={end_date.date()}, End Price={end_price:.2f}, Actual Years={actual_years:.2f}")

                 is_end_valid = pd.notna(end_price) and math.isfinite(end_price)
                 is_start_valid = pd.notna(start_price) and math.isfinite(start_price) and start_price > 1e-9
                 # Check if period is reasonably close to target years
                 is_period_valid = abs(actual_years - num_years_cagr) < 0.5 and actual_years > 0

                 if is_end_valid and is_start_valid and is_period_valid:
                      stock_cagr = calculate_cagr_modified(start_price, end_price, actual_years)
                      if pd.notna(stock_cagr) and math.isfinite(stock_cagr):
                          calculated_cagrs['Stock'] = stock_cagr; valid_cagr_list.append(stock_cagr)
                          projection_log.append(f"  ✅ Stock CAGR:   {format_percent(stock_cagr)} (Using '{price_col}')")
                      else: projection_log.append(f"  ❌ Stock CAGR Calculation Failed (Result Invalid: {stock_cagr}).")
                 else:
                      reason = [];
                      if not is_end_valid: reason.append("Invalid End Price");
                      if not is_start_valid: reason.append(f"Invalid/Non-positive Start Price ({start_price:.2f})")
                      if not is_period_valid: reason.append(f"Period Invalid ({actual_years:.2f} yrs vs target {num_years_cagr})")
                      projection_log.append(f"  ❌ Stock CAGR Skipped ({', '.join(reason)}).")
             except Exception as e: projection_log.append(f"  ❌ Stock CAGR Skipped (Error during lookup/calculation: {e})")
         elif not price_col: pass # Already logged error
         else: projection_log.append(f"  ❌ Stock CAGR Skipped (History index is not DatetimeIndex)")
    else:
         reason = [];
         if history_long_df is None: reason.append("History DF is None")
         elif history_long_df.empty: reason.append("History DF is empty")
         elif len(history_long_df) < 2: reason.append("Less than 2 data points")
         projection_log.append(f"  ❌ Stock CAGR Skipped (Insufficient/invalid history data object: {', '.join(reason)})")

    # 4. Calculate Geometric Mean of the valid CAGRs
    projection_log.append("\nCalculating Geometric Mean of Valid CAGRs...")
    if valid_cagr_list:
        # Check for extreme outliers before GM (e.g., > 500% or < -90%) which might skew GM
        reasonable_cagr_list = [r for r in valid_cagr_list if -0.90 <= r <= 5.0]
        if len(reasonable_cagr_list) < len(valid_cagr_list):
             projection_log.append(f"  ⚠️ Excluding {len(valid_cagr_list) - len(reasonable_cagr_list)} extreme CAGR(s) from GM calculation.")
             if not reasonable_cagr_list: # All were extreme
                 projection_log.append(f"  ⚠️ All valid CAGRs were extreme. Falling back to default growth.")
                 initial_growth_rate = default_init_growth; growth_source = f"Default ({format_percent(default_init_growth)}) [All CAGRs Extreme]"
             else: # Use only the reasonable ones
                 gm_growth = geometric_mean_modified(reasonable_cagr_list)
                 if pd.notna(gm_growth) and math.isfinite(gm_growth):
                    initial_growth_rate = gm_growth; growth_source = f"GM of {len(reasonable_cagr_list)} reasonable CAGR(s)"
                    projection_log.append(f"  ▶️ Calculated Geometric Mean (Reasonable Only) = {format_percent(initial_growth_rate)}")
                    included_cagrs = [f"{k}:{format_percent(v)}" for k, v in calculated_cagrs.items() if v in reasonable_cagr_list]
                    projection_log.append(f"     Based on: [{', '.join(included_cagrs)}]")
                 else:
                     projection_log.append(f"  ⚠️ Geometric Mean calculation failed even on reasonable CAGRs. Falling back to default growth.")
                     initial_growth_rate = default_init_growth; growth_source = f"Default ({format_percent(default_init_growth)}) [GM Fail]"
                     projection_log.append(f"     Reasonable CAGRs: {reasonable_cagr_list}")
        else: # No extreme outliers detected
             gm_growth = geometric_mean_modified(valid_cagr_list)
             if pd.notna(gm_growth) and math.isfinite(gm_growth):
                initial_growth_rate = gm_growth; growth_source = f"GM of {len(valid_cagr_list)} valid CAGR(s)"
                projection_log.append(f"  ▶️ Calculated Geometric Mean = {format_percent(initial_growth_rate)}")
                included_cagrs = [f"{k}:{format_percent(v)}" for k, v in calculated_cagrs.items()]
                projection_log.append(f"     Based on: [{', '.join(included_cagrs)}]")
             else:
                projection_log.append(f"  ⚠️ Geometric Mean calculation failed or result invalid. Falling back to default growth.")
                initial_growth_rate = default_init_growth; growth_source = f"Default ({format_percent(default_init_growth)}) [GM Fail]"
                projection_log.append(f"     CAGRs calculated but GM failed: {valid_cagr_list}")
    else:
        projection_log.append(f"  ⚠️ No valid 3yr CAGRs could be calculated. Falling back to default growth.")
        initial_growth_rate = default_init_growth; growth_source = f"Default ({format_percent(default_init_growth)}) [No Valid CAGRs]"

    # 5. Apply Floor (1.1%) and Ceiling (15%) to Initial Growth Rate
    unbounded_rate = initial_growth_rate
    initial_growth_rate = max(initial_growth_rate, MIN_INITIAL_GROWTH_FLOOR); initial_growth_rate = min(initial_growth_rate, MAX_INITIAL_GROWTH_CEILING)
    if abs(initial_growth_rate - unbounded_rate) > 1e-6:
        growth_source += f" (Bounded from {format_percent(unbounded_rate)})"
        projection_log.append(f"  Applied Bounds: Initial Growth Rate = {format_percent(initial_growth_rate)} (Original: {format_percent(unbounded_rate)}, Floor: {format_percent(MIN_INITIAL_GROWTH_FLOOR)}, Ceiling: {format_percent(MAX_INITIAL_GROWTH_CEILING)})")
    else:
        projection_log.append(f"  Final Initial Growth Rate = {format_percent(initial_growth_rate)} (Floor: {format_percent(MIN_INITIAL_GROWTH_FLOOR)}, Ceiling: {format_percent(MAX_INITIAL_GROWTH_CEILING)})")

    projection_results['initial_growth_rate'] = initial_growth_rate; projection_results['growth_source'] = growth_source

    # ==============================================================
    # PHASE 4: Determine Terminal Growth & Project FCFF
    # ==============================================================
    projection_log.append("\n--- Phase 4: Determining Terminal Growth & Projecting FCFF ---")

    # --- Terminal Growth Rate Calculation ---
    g_cap_initial_final = initial_growth_rate if pd.notna(initial_growth_rate) and math.isfinite(initial_growth_rate) else TERMINAL_GROWTH_RATE_CAP
    g_cap_wacc = disc_rate * 0.60 # Conservative WACC cap
    g_cap_rf = final_user_rf * 0.90 # Conservative Rf cap
    g_cap_config = TERMINAL_GROWTH_RATE_CAP

    projection_log.append(f"\nDetermining Terminal Growth Rate...")
    projection_log.append(f"  Caps Considered: Config={format_percent(g_cap_config)}, WACC*0.60={format_percent(g_cap_wacc)}, Initial(Final)={format_percent(g_cap_initial_final)}, Rf*0.90={format_percent(g_cap_rf)}")

    # Find minimum of *valid* caps only
    valid_caps = [cap for cap in [g_cap_config, g_cap_wacc, g_cap_initial_final, g_cap_rf] if np.isfinite(cap)]
    if not valid_caps:
        projection_log.append("  ⚠️ ERROR: No valid caps for terminal growth. Using 0%.")
        terminal_growth_rate = 0.0
    else:
        terminal_growth_rate = min(valid_caps); projection_log.append(f"  ↳ Minimum of Valid Caps: {format_percent(terminal_growth_rate)}")

    original_g_before_floor = terminal_growth_rate
    # Ensure terminal growth isn't negative (apply floor)
    terminal_growth_rate = max(MIN_GROWTH, terminal_growth_rate)
    if abs(terminal_growth_rate - original_g_before_floor) > 1e-6 : projection_log.append(f"  ↳ Adjusted by Floor ({format_percent(MIN_GROWTH)}): {format_percent(terminal_growth_rate)}")

    # Ensure terminal growth is strictly less than WACC
    if terminal_growth_rate >= disc_rate:
        original_g = terminal_growth_rate;
        terminal_growth_rate = disc_rate - 0.005 # Reduce slightly below WACC
        terminal_growth_rate = max(MIN_GROWTH, terminal_growth_rate) # Ensure floor still holds
        projection_log.append(f"  ⚠️ Terminal Growth ({format_percent(original_g)}) >= WACC ({format_percent(disc_rate)}). Reduced to {format_percent(terminal_growth_rate)}.")

    projection_log.append(f"  ▶️ Using Final Terminal Growth Rate (g): {format_percent(terminal_growth_rate)}"); projection_results['terminal_growth_rate'] = terminal_growth_rate

    # --- Project FCFFs ---
    projected_fcffs = []; growth_rates_applied = []; pv_fcffs = []; cumulative_pv_fcff = 0.0
    current_fcff_for_proj = fcff_start_basis # Use the basis chosen/derived in Phase 2

    # Ensure initial growth is valid, default to 0 otherwise (should be handled by bounds, but extra check)
    if pd.isna(initial_growth_rate) or not math.isfinite(initial_growth_rate):
        projection_log.append(f"⚠️ FATAL: Invalid final initial growth rate ({initial_growth_rate}) despite checks. Using 0% for projection start.")
        effective_initial_growth = 0.0
    else: effective_initial_growth = initial_growth_rate

    # Ensure terminal growth is valid, default to 0 otherwise (should be handled by bounds/WACC check)
    if pd.isna(terminal_growth_rate) or not math.isfinite(terminal_growth_rate):
         projection_log.append(f"⚠️ FATAL: Invalid final terminal growth rate ({terminal_growth_rate}) despite checks. Using 0% for terminal growth.")
         effective_terminal_growth = 0.0
    else: effective_terminal_growth = terminal_growth_rate

    projection_log.append("\n--- FCFF Projections ---");
    projection_log.append(f"Starting Basis (T=0): {format_currency(current_fcff_for_proj, '', 0, is_indian=is_indian)} (Method: {fcff_basis_method})")
    projection_log.append(f"{'Year':<5} {'Growth':<10} {'FCFF':<18} {'PV Factor':<10} {'PV FCFF':<18}"); projection_log.append("-" * 65)
    last_valid_fcff = current_fcff_for_proj

    for year in range(1, forecast_yrs + 1):
        # Linear decay of growth rate from initial to terminal
        if forecast_yrs > 0: decay_factor = max(0, (forecast_yrs - year)) / forecast_yrs
        else: decay_factor = 0
        growth_rate = effective_terminal_growth + (effective_initial_growth - effective_terminal_growth) * decay_factor
        growth_rate = max(MIN_GROWTH, growth_rate) # Ensure growth rate doesn't go below floor

        growth_rates_applied.append(growth_rate)
        fcff_forecast = last_valid_fcff * (1 + growth_rate)

        # Handle potential non-finite results robustly
        if not np.isfinite(fcff_forecast):
             projection_log.append(f"⚠️ Year {year} FCFF forecast became non-finite (Start:{last_valid_fcff}, Growth:{growth_rate}). Setting to 0.")
             fcff_forecast = 0.0
        elif abs(fcff_forecast) > 1e18: # Check for excessive magnitude
             projection_log.append(f"⚠️ Year {year} FCFF forecast ({format_currency(fcff_forecast,'',0)}) extremely large. Capping/setting to 0.")
             fcff_forecast = 0.0

        projected_fcffs.append(fcff_forecast)
        last_valid_fcff = fcff_forecast # Update for next iteration

        pv_factor = 1 / ((1 + disc_rate) ** year); pv_fcff = fcff_forecast * pv_factor
        if not np.isfinite(pv_fcff): pv_fcff = 0.0 # Set PV to 0 if non-finite
        pv_fcffs.append(pv_fcff); cumulative_pv_fcff += pv_fcff
        projection_log.append(f"{year:<5} {format_percent(growth_rate):<10} {format_currency(fcff_forecast, '', 0, is_indian=is_indian):<18} {pv_factor:<10.4f} {format_currency(pv_fcff, '', 0, is_indian=is_indian):<18}")

    projection_results['projected_fcffs'] = projected_fcffs; projection_results['growth_rates'] = growth_rates_applied; projection_results['pv_fcffs'] = pv_fcffs
    projection_results['cumulative_pv_fcff'] = cumulative_pv_fcff if np.isfinite(cumulative_pv_fcff) else 0.0 # Ensure finite
    projection_log.append("-" * 65); projection_log.append(f"Sum of PV(Forecasted FCFFs): {format_currency(projection_results['cumulative_pv_fcff'], symbol=currency_sym, is_indian=is_indian)}")

    # --- Calculate Terminal Value ---
    projection_log.append("\n--- Terminal Value Calculation ---")
    terminal_value = np.nan; pv_terminal_value = np.nan
    final_forecast_fcff = projected_fcffs[-1] if projected_fcffs else np.nan

    # Check if essential components for TV calc are valid
    if (disc_rate - effective_terminal_growth) <= 1e-6: # Denominator check (WACC > g)
         projection_log.append(f"❌ ERROR: WACC ({format_percent(disc_rate)}) <= Terminal Growth ({format_percent(effective_terminal_growth)}). TV calculation failed."); projection_results['error'] = "WACC <= g. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    elif not np.isfinite(final_forecast_fcff):
         projection_log.append(f"❌ ERROR: Final projected FCFF (Year {forecast_yrs}) is invalid ({final_forecast_fcff}). TV calculation failed."); projection_results['error'] = "Final FCFF invalid. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    else:
         # Proceed with calculation only if components are valid
         fcff_n_plus_1 = final_forecast_fcff * (1 + effective_terminal_growth)
         terminal_value = fcff_n_plus_1 / (disc_rate - effective_terminal_growth)

         # Check if calculated TV is valid (finite and non-negative)
         if not np.isfinite(terminal_value) or terminal_value < 0:
              projection_log.append(f"⚠️ Calculated TV invalid/negative ({format_currency(terminal_value, currency_sym, is_indian=is_indian)}). Final FCFF={format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)}, g={format_percent(effective_terminal_growth)}, WACC={format_percent(disc_rate)}. Setting TV=0.")
              terminal_value = 0.0; pv_terminal_value = 0.0
              if projection_results['error'] is None: projection_results['error'] = "Calculated Terminal Value invalid/negative." # Add non-fatal error
         else:
              # Calculate PV of Terminal Value
              pv_factor_tv = 1 / ((1 + disc_rate) ** forecast_yrs); pv_terminal_value = terminal_value * pv_factor_tv
              if not np.isfinite(pv_terminal_value):
                    projection_log.append(f"⚠️ PV(TV) calculation failed (PV Factor: {pv_factor_tv:.4f}). Setting PV(TV)=0.")
                    pv_terminal_value = 0.0
              projection_log.append(f"FCFF Yr {forecast_yrs}: {format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)}")
              projection_log.append(f"FCFF Yr {forecast_yrs+1} (Projected): {format_currency(fcff_n_plus_1, currency_sym, is_indian=is_indian)}")
              projection_log.append(f"Terminal Value (at Yr {forecast_yrs}): {format_currency(terminal_value, currency_sym, is_indian=is_indian)}")
              projection_log.append(f"PV Factor TV (Yr {forecast_yrs}): {pv_factor_tv:.4f} -> PV(TV): {format_currency(pv_terminal_value, currency_sym, is_indian=is_indian)}")

    projection_results['terminal_value'] = terminal_value if np.isfinite(terminal_value) else 0.0
    projection_results['pv_terminal_value'] = pv_terminal_value if np.isfinite(pv_terminal_value) else 0.0

    # Display the full log at the end
    display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
    return projection_results

# --- Execute FCFF calculation and projection ---
# Initialize fcff_results dictionary to ensure it's defined before the try block
fcff_results = {'error': 'Calculation not run yet.', 'initial_growth_rate': np.nan, 'growth_source':'N/A', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'N/A'}

if np.isfinite(discount_rate):
    try:
        # Use the V21.2 (User Choice) function name
        fcff_results = calculate_and_project_fcff_v21_user_choice(
            stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'], stock_data['history_long'], # Pass history_long
            wacc_components, discount_rate, FORECAST_YEARS, risk_free_rate_input, DEFAULT_GROWTH_RATE,
            currency_symbol, is_indian_stock)
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during FCFF Projection: {e}**"))
         print(traceback.format_exc()) # Print detailed traceback
         # Ensure fcff_results is still a dict with error info
         fcff_results = {'error': f'Runtime error in FCFF calc: {e}', 'initial_growth_rate': np.nan, 'growth_source':'Error', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'Error'}

else:
    display(Markdown("❌ **Skipping FCFF Projection due to invalid WACC.**"));
    # Ensure fcff_results is a dict even if skipped
    fcff_results = {'error': 'Invalid WACC.', 'initial_growth_rate': np.nan, 'growth_source':'N/A (WACC Invalid)', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'N/A'}


# %% [markdown]
# ## 7. Enterprise Value and Equity Value Calculation
# (No changes needed in this section's logic based on the request, keeping previous robust version)

# %%
# Valuation Function Definition
def calculate_final_valuation(fcff_data, balance_sheet_df, shares_outstanding_val, currency_sym, wacc_data, is_indian):
    display(Markdown("### Final Valuation Calculation..."))
    valuation = {'enterprise_value': np.nan, 'equity_value': np.nan, 'intrinsic_value_per_share': np.nan, 'error': None}
    valuation_log = []

    # Ensure fcff_data is a dictionary before accessing keys
    if not isinstance(fcff_data, dict):
        error_msg = f"FCFF Data is not a dictionary (type: {type(fcff_data)}). Cannot proceed."
        valuation['error'] = error_msg; valuation_log.append(f"❌ ERROR: {error_msg}")
        display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    # Check for fatal FCFF error preventing projection
    fcff_fatal_error = fcff_data.get('error') and "Cannot proceed" in fcff_data.get('error', '')
    if fcff_fatal_error:
         valuation['error'] = f"FCFF Error: {fcff_data.get('error', 'Unknown')}"; valuation_log.append(f"❌ ERROR: Upstream FCFF error prevents valuation: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    elif fcff_data.get('error'): # Non-fatal error (e.g., TV calc failed)
         valuation_log.append(f"ℹ️ Proceeding with EV calculation despite FCFF warning: {fcff_data.get('error')}")

    pv_forecast_sum = fcff_data.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_data.get('pv_terminal_value', np.nan)

    # Check if components are valid numbers (allow zero), use 0 if invalid
    if not np.isfinite(pv_forecast_sum): pv_forecast_sum = 0.0; valuation_log.append(f"⚠️ PV(Forecast FCFFs) invalid, using 0.")
    if not np.isfinite(pv_tv): pv_tv = 0.0; valuation_log.append(f"⚠️ PV(Terminal Value) invalid, using 0.")

    enterprise_value = pv_forecast_sum + pv_tv; valuation['enterprise_value'] = enterprise_value
    valuation_log.append(f"PV(Forecast FCFFs): {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"PV(Terminal Value):   {format_currency(pv_tv, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"↳ Enterprise Value (EV): {format_currency(enterprise_value, currency_sym, is_indian=is_indian)}")

    # Bridge EV to Equity Value
    total_debt_latest = 0.0; cash_and_equivalents = 0.0; minority_interest = 0.0
    debt_source_info = '?';
    # Get consistent debt value from WACC calculation
    if isinstance(wacc_data, dict):
        debt_source_info = wacc_data.get('debt_calc_method', '?');
        total_debt_latest = wacc_data.get('debt_value_used', 0.0)
        if not np.isfinite(total_debt_latest): total_debt_latest = 0.0 # Ensure numeric
    else: valuation_log.append(f"⚠️ WACC data invalid, cannot get debt used in WACC. Assuming 0 for bridge.")

    # Get Cash & Minority Interest from latest Balance Sheet
    if balance_sheet_df is not None and not balance_sheet_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]
        valuation_log.append(f"\nBridging EV to Equity:")
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        cash_keys = ["Cash And Cash Equivalents", "Cash Cash Equivalents And Short Term Investments", "Cash Financial", "Cash", "Cash And Short Term Investments"];
        cash_and_equivalents = safe_get(latest_bs, cash_keys, "BS (Cash)", default=0)
        if not np.isfinite(cash_and_equivalents): cash_and_equivalents = 0.0 # Ensure numeric
        valuation_log.append(f"  + Cash & Equiv (BS): {format_currency(cash_and_equivalents, currency_sym, is_indian=is_indian)}")

        # Check for Minority Interest / Noncontrolling Interest
        minority_interest_keys = ["Minority Interest", "Noncontrolling Interest"]
        minority_interest = safe_get(latest_bs, minority_interest_keys, "BS (Minority Int)", default=0)
        if not np.isfinite(minority_interest): minority_interest = 0.0 # Ensure numeric
        if abs(minority_interest) > 1e-6: # Only log if non-zero
             valuation_log.append(f"  - Minority Interest (BS): {format_currency(minority_interest, currency_sym, is_indian=is_indian)}")
    else:
        valuation_log.append("⚠️ Balance Sheet empty or invalid. Cannot fetch Cash/Minority Interest for EV bridge.")
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        valuation_log.append(f"  + Cash & Equiv: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")
        valuation_log.append(f"  - Minority Interest: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")

    # Check components before calculating Equity Value
    if not all(np.isfinite(v) for v in [enterprise_value, total_debt_latest, cash_and_equivalents, minority_interest]):
         valuation['error'] = "Invalid component(s) for EV->Equity bridge.";
         valuation_log.append(f"❌ ERROR: {valuation['error']} EV={enterprise_value}, Debt={total_debt_latest}, Cash={cash_and_equivalents}, MI={minority_interest}")
         valuation['equity_value'] = np.nan
         valuation['intrinsic_value_per_share'] = np.nan
         display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation

    equity_value = enterprise_value - total_debt_latest + cash_and_equivalents - minority_interest;
    valuation['equity_value'] = equity_value
    valuation_log.append(f"↳ Equity Value: {format_currency(equity_value, currency_sym, is_indian=is_indian)}")

    # Calculate Intrinsic Value Per Share
    intrinsic_value_per_share = np.nan
    if not np.isfinite(equity_value):
        valuation['error'] = f"Equity value invalid ({equity_value}). Cannot calculate Per Share value.";
        valuation_log.append(f"❌ ERROR: {valuation['error']}")
    elif shares_outstanding_val <= 0:
        valuation['error'] = f"Shares outstanding invalid ({shares_outstanding_val}). Cannot calculate Per Share value.";
        valuation_log.append(f"❌ ERROR: {valuation['error']}")
    else:
        intrinsic_value_per_share = equity_value / shares_outstanding_val;
        valuation_log.append(f"  ÷ Shares Out: {shares_outstanding_val:,.0f}")
        valuation_log.append(f"▶️ Intrinsic Value/Share: {format_currency(intrinsic_value_per_share, currency_sym, is_indian=is_indian)}")

    valuation['intrinsic_value_per_share'] = intrinsic_value_per_share if np.isfinite(intrinsic_value_per_share) else np.nan
    display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
    return valuation

# --- Execute Final Valuation ---
final_valuation = {'error': 'Calculation not run yet.', 'intrinsic_value_per_share': np.nan}
# Check if FCFF results exist and are a dict, WACC components exist and are dict, and FCFF didn't have a *fatal* error stopping projection
fcff_error_is_fatal = isinstance(fcff_results, dict) and fcff_results.get('error') and "Cannot proceed" in fcff_results.get('error', '')

if isinstance(fcff_results, dict) and isinstance(wacc_components, dict) and not fcff_error_is_fatal:
    try:
        final_valuation = calculate_final_valuation(
            fcff_results,
            stock_data['balance_sheet'],
            shares_outstanding, # Use the potentially corrected value
            currency_symbol,
            wacc_components, # Pass WACC data for consistent debt
            is_indian_stock
        )
    except Exception as e:
        display(Markdown(f"❌ **CRITICAL ERROR during Final Valuation: {e}**"))
        print(traceback.format_exc())
        final_valuation = {'error': f'Runtime error in final valuation: {e}', 'intrinsic_value_per_share': np.nan}
elif isinstance(fcff_results, dict) and fcff_error_is_fatal:
     display(Markdown(f"❌ **Skipping Final Valuation due to fatal FCFF error:** {fcff_results.get('error')}"));
     final_valuation = {'error': fcff_results.get('error', 'Upstream fatal FCFF error.'), 'intrinsic_value_per_share': np.nan}
else:
     display(Markdown("❌ **Skipping Final Valuation due to critical upstream errors (invalid WACC/FCFF data structures).**"));
     final_valuation = {'error': 'FCFF or WACC results dictionary not created/valid.', 'intrinsic_value_per_share': np.nan}


# %% [markdown]
# ## 8. Financial Ratio Analysis (Ratio Plot Fixes V21.2)

# %%
# Ratio Function Definition (Ratio Plot Fixes V21.2)
def analyze_ratios(stock_info_dict, financials_df, balance_sheet_df, cashflow_df,
                   enterprise_val, current_price_val, shares_outstanding_val,
                   currency_sym, user_benchmark_dict, # Benchmarks stored as entered (e.g., 15.0 for 15%)
                   benchmark_source_dict, # Dict tracking source per ratio
                   wacc_data, is_indian): # Pass WACC data for consistent debt and flag

    """ Calculates ratios, compares to user/fallback benchmarks, uses WACC-consistent debt. Plots fixed. """
    display(Markdown("### Financial Ratio Analysis..."))
    ratios = {} # Calculated ratios {RatioName: value (decimal form)}
    ratio_log = []
    ratio_data_for_table = [] # For HTML table display
    ratio_plotly_data = {'Ratio': [], 'Company': [], 'Benchmark': [], 'Benchmark Source': [], 'Status': []} # For Plotly chart
    company_nm = stock_info_dict.get("shortName", ticker_symbol)

    # --- Pre-checks ---
    valid_price = current_price_val is not None and np.isfinite(current_price_val) and current_price_val > 0
    valid_shares = shares_outstanding_val is not None and np.isfinite(shares_outstanding_val) and shares_outstanding_val > 0
    valid_enterprise_val = enterprise_val is not None and np.isfinite(enterprise_val) # EV can be negative

    if not valid_price: ratio_log.append("⚠️ Current Price invalid/unavailable. Price-based ratios (P/E, P/B) will be N/A.")
    if not valid_shares: ratio_log.append("⚠️ Shares Outstanding invalid/unavailable. Per-share ratios (EPS, BVPS, P/E, P/B) will be N/A.")
    if not valid_enterprise_val: ratio_log.append("⚠️ Enterprise Value invalid/unavailable. EV/EBITDA will be N/A.")


    # Check for essential statements
    essential_statements_missing = False
    if financials_df is None or financials_df.empty:
         ratio_log.append("❌ ERROR: Income Statement empty. Most ratios cannot be calculated.")
         essential_statements_missing = True
    if balance_sheet_df is None or balance_sheet_df.empty:
         ratio_log.append("❌ ERROR: Balance Sheet empty. Many ratios cannot be calculated.")
         essential_statements_missing = True
    if cashflow_df is None or cashflow_df.empty:
         ratio_log.append("⚠️ Warning: Cash Flow statement empty. EBITDA calculation might be inaccurate (assuming 0 Depreciation).")

    if essential_statements_missing:
         display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
         return {}, [] # Return empty results

    # Get latest columns safely
    latest_fin = financials_df.iloc[:, 0] if not financials_df.empty else pd.Series(dtype=float)
    latest_bs = balance_sheet_df.iloc[:, 0] if not balance_sheet_df.empty else pd.Series(dtype=float)
    latest_cf = cashflow_df.iloc[:, 0] if not cashflow_df.empty else pd.Series(dtype=float)

    # --- Calculate Ratio Components (using safe_get with NaN default where appropriate) ---
    # Values are stored in their natural scale (e.g., millions, billions)
    net_income = safe_get(latest_fin, ["Net Income Continuous Operations", "Net Income", "Net Income Common Stockholders", "Net Income Applicable To Common Shares"], default=np.nan)
    revenue = safe_get(latest_fin, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)

    # Depreciation (Robust fetching)
    depreciation = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], "Cash Flow (Dep)", default=np.nan)
    if not np.isfinite(depreciation):
        dep_is = safe_get(latest_fin, ["Depreciation And Amortization", "Depreciation Income Statement"], "Income Stmt (Dep Fallback)", default=np.nan)
        if np.isfinite(dep_is): depreciation = dep_is; ratio_log.append("ℹ️ Used Depreciation from Income Statement (CF unavailable/invalid).")
        else: depreciation = 0.0; ratio_log.append("⚠️ Depreciation not found in CF or IS. Assuming 0 for EBITDA calc.")
    else: depreciation = float(depreciation)

    # EBITDA (Calculated robustly)
    ebitda = np.nan
    if np.isfinite(ebit) and np.isfinite(depreciation): ebitda = ebit + depreciation
    elif np.isfinite(ebit): ebitda = ebit; ratio_log.append("ℹ️ Calculating EBITDA using EBIT only (Depreciation unavailable/zero).")
    else: ratio_log.append("⚠️ EBIT invalid. Cannot calculate EBITDA.")

    # Balance Sheet items
    total_equity = safe_get(latest_bs, ["Total Stockholder Equity", "Stockholders Equity", "Total Equity Gross Minority Interest", "Common Stock Equity"], "BS (Equity)", default=np.nan)
    total_assets = safe_get(latest_bs, ["Total Assets"], "BS (Assets)", default=np.nan)
    current_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], "BS (Curr Assets)", default=np.nan)
    current_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], "BS (Curr Liab)", default=np.nan)
    inventory = safe_get(latest_bs, ["Inventory"], "BS (Inventory)", default=np.nan)
    if not np.isfinite(inventory): inventory = 0.0; ratio_log.append("ℹ️ Inventory not found on BS. Assuming 0 for Quick Ratio calc.")
    else: inventory = float(inventory)

    # Use Debt value consistent with WACC calculation
    total_debt_ratio = 0.0; debt_source_info = '?'
    if isinstance(wacc_data, dict):
        total_debt_ratio = wacc_data.get('debt_value_used', 0.0)
        debt_source_info = wacc_data.get('debt_calc_method', '?');
        if not np.isfinite(total_debt_ratio): total_debt_ratio = 0.0
    else: ratio_log.append(f"⚠️ WACC data invalid, cannot get consistent debt. Assuming 0 for ratio calcs."); total_debt_ratio = 0.0
    ratio_log.append(f"Using Debt = {format_currency(total_debt_ratio, currency_sym, is_indian=is_indian)} (Source: WACC - {debt_source_info}) for relevant ratios.")

    # Per Share Metrics
    eps = (net_income / shares_outstanding_val) if valid_shares and np.isfinite(net_income) and net_income is not None else np.nan
    bvps = (total_equity / shares_outstanding_val) if valid_shares and np.isfinite(total_equity) and total_equity is not None else np.nan

    # Capital Employed (ROCE denominator) - Robust calculation
    capital_employed = np.nan; ce_source = "N/A"
    # Definition 1: Total Assets - Current Liabilities
    if np.isfinite(total_assets) and np.isfinite(current_liab):
        capital_employed_def1 = total_assets - current_liab
        if np.isfinite(capital_employed_def1) and capital_employed_def1 > 1.0: capital_employed = capital_employed_def1; ce_source = "Total Assets - Current Liab"
    # Definition 2 (Fallback): Total Equity + Total Debt
    if not np.isfinite(capital_employed): # If Def1 failed or yielded non-positive
        if np.isfinite(total_equity) and np.isfinite(total_debt_ratio):
             capital_employed_def2 = total_equity + total_debt_ratio
             if np.isfinite(capital_employed_def2) and capital_employed_def2 > 1.0: capital_employed = capital_employed_def2; ce_source = "Total Equity + Total Debt"
    if not np.isfinite(capital_employed): ce_source = "Invalid/Unavailable"; ratio_log.append("⚠️ Capital Employed could not be reliably calculated (needed for ROCE). Check BS data.")

    # --- Calculate Ratios (Result is ratio value, e.g., 0.15 for 15% ROE) ---
    def safe_division(numerator, denominator):
        # Ensure inputs are numeric before check
        num_val = float(numerator) if np.isfinite(numerator) else np.nan
        den_val = float(denominator) if np.isfinite(denominator) else np.nan
        if not np.isfinite(num_val) or not np.isfinite(den_val): return np.nan
        if abs(den_val) < 1e-9: return np.nan # Avoid division by zero
        try: return num_val / den_val
        except: return np.nan

    # Store ratios as decimals where applicable (e.g., ROE = 0.15)
    ratios['P/E'] = safe_division(current_price_val, eps) if valid_price else np.nan
    ratios['P/B'] = safe_division(current_price_val, bvps) if valid_price else np.nan
    ratios['EV/EBITDA'] = safe_division(enterprise_val, ebitda) if valid_enterprise_val else np.nan
    ratios['ROE'] = safe_division(net_income, total_equity)
    ratios['ROA'] = safe_division(net_income, total_assets)
    ratios['ROCE'] = safe_division(ebit, capital_employed)
    ratios['Asset Turnover'] = safe_division(revenue, total_assets)
    ratios['Current Ratio'] = safe_division(current_assets, current_liab)
    quick_assets = (current_assets - inventory) if np.isfinite(current_assets) and np.isfinite(inventory) else np.nan
    ratios['Quick Ratio'] = safe_division(quick_assets, current_liab)

    # --- Compare with User/Fallback Benchmarks & Prepare Data Tables ---
    display(Markdown("*(Comparing ratios to user/fallback benchmarks...)*"))
    for key in REQUIRED_BENCHMARK_RATIOS:
        # company_value is the calculated ratio (decimal for percentages)
        company_value = ratios.get(key)
        is_percent_type = key in ['ROE', 'ROA', 'ROCE']

        # Format company value for display
        value_fmt = "N/A"
        if company_value is None or pd.isna(company_value): value_fmt = "N/A"
        elif not math.isfinite(company_value): value_fmt = "Inf" if company_value > 0 else "-Inf"
        else: # Format finite values
            if is_percent_type: value_fmt = format_percent(company_value, 1) # e.g., 15.0%
            elif key == 'Asset Turnover': value_fmt = f"{company_value:.2f}" # Not %
            else: value_fmt = format_currency(company_value, '', 2, is_indian=False) # Standard number format

        # Get benchmark (stored as number, e.g., 19.05 for ROE benchmark)
        user_benchmark_val = user_benchmark_dict.get(key, None)
        fallback_benchmark_val = DEFAULT_BENCHMARK_RATIOS.get(key, None)
        benchmark_source = benchmark_source_dict.get(key, "?")
        final_benchmark_val = None # This will be the value used for comparison (e.g., 19.05)

        if benchmark_source == 'User Input' and user_benchmark_val is not None and np.isfinite(user_benchmark_val):
            final_benchmark_val = user_benchmark_val
        elif benchmark_source == 'Fallback Default' and fallback_benchmark_val is not None and np.isfinite(fallback_benchmark_val):
            final_benchmark_val = fallback_benchmark_val
        else:
            benchmark_source = "Not Available"

        # Format benchmark for display
        bench_fmt = "N/A"
        if final_benchmark_val is not None:
             if is_percent_type: bench_fmt = f"{final_benchmark_val:.1f}%" # Format as %
             elif key == 'Asset Turnover': bench_fmt = f"{final_benchmark_val:.2f}"
             else: bench_fmt = f"{final_benchmark_val:.2f}" # Standard number format
             bench_fmt += f" <i style='font-size:0.8em; color:grey'>({benchmark_source.replace('Fallback ','')})</i>"
        else: bench_fmt = "<i style='color:grey'>Not Provided</i>"

        # Compare and determine status
        status = "<i style='color:grey'>N/A</i>"; status_color = "grey"
        # Comparison requires valid finite company value and benchmark value
        if np.isfinite(company_value) and final_benchmark_val is not None and np.isfinite(final_benchmark_val):
            # For comparison, convert company value to percentage if needed
            numeric_value_comp = company_value * 100 if is_percent_type else company_value
            numeric_benchmark_comp = final_benchmark_val # Benchmark is already stored as % value (e.g., 19.05)
            # Asset turnover benchmark is not percentage
            if key == 'Asset Turnover': numeric_benchmark_comp = final_benchmark_val

            good_color = '#2ecc71'; bad_color = '#e74c3c'; neutral_color = '#f39c12'; grey_color='#7f8c8d'

            # Calculate relative difference vs benchmark, handle zero benchmark
            if abs(numeric_benchmark_comp) > 1e-9:
                 ratio_diff = (numeric_value_comp - numeric_benchmark_comp) / abs(numeric_benchmark_comp)
            else: # Handle benchmark near zero
                 ratio_diff = float('inf') if numeric_value_comp > 0 else float('-inf') if numeric_value_comp < 0 else 0.0

            # Determine status based on ratio type
            # Profitability/Efficiency: Higher is better
            if key in ['ROE', 'ROA', 'ROCE', 'Asset Turnover', 'Current Ratio', 'Quick Ratio']:
                 if ratio_diff > 0.20: status_color = good_color; status = f"⬆️ Strong"
                 elif ratio_diff < -0.20: status_color = bad_color; status = f"⬇️ Weak"
                 else: status_color = grey_color; status = f"~ Average"
            # Valuation Multiples: Lower often better/cheaper, but context matters
            elif key in ['P/E', 'P/B', 'EV/EBITDA']:
                 if ratio_diff < -0.20: status_color = good_color; status = f"✅ Low (vs Bench)"
                 elif ratio_diff > 0.50: status_color = bad_color; status = f"‼️ High (vs Bench)"
                 elif ratio_diff > 0.20: status_color = neutral_color; status = f"⚠️ Elevated (vs Bench)"
                 else: status_color = grey_color; status = f"~ Average (vs Bench)"
            status = f"<span style='color:{status_color}'>{status}</span>"

            # Prepare data for Plotly chart
            # Use percentage values for both company and benchmark if it's a percent type
            plot_value = numeric_value_comp
            plot_benchmark = numeric_benchmark_comp
            # Append data for plotting
            ratio_plotly_data['Ratio'].append(key);
            ratio_plotly_data['Company'].append(plot_value);
            ratio_plotly_data['Benchmark'].append(plot_benchmark);
            ratio_plotly_data['Benchmark Source'].append(benchmark_source);
            ratio_plotly_data['Status'].append(status) # Status string for potential use
        elif not np.isfinite(company_value): status = "<i style='color:grey'>Company N/A</i>"
        elif final_benchmark_val is None: status = "<i style='color:grey'>No Benchmark</i>"
        # Add row to HTML table data
        ratio_data_for_table.append([f"<b>{key}</b>", value_fmt, bench_fmt, status])

    # --- Display Logs & Table ---
    if ratio_log: display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
    if ratio_data_for_table:
        header = ["Ratio", f"{company_nm} Value", "Benchmark (Source)", "Status"]; html_table = "<p>(*Status vs User/Default benchmarks. Threshold: +/- 20% approx. Valuation multiples judged vs benchmark.)</p>"; html_table += "<table border='1' style='border-collapse:collapse; width:98%; margin-left:auto; margin-right:auto; font-size: 0.9em;'>"; html_table += "<thead style='background-color:#e9ecef; color: #495057;'><tr>" + "".join(f"<th style='padding: 8px; text-align:center;'>{h}</th>" for h in header) + "</tr></thead>"; html_table += "<tbody>"
        for i, row in enumerate(ratio_data_for_table): bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff"; html_table += f"<tr style='background-color:{bg_color};'>"; html_table += f"<td style='padding: 6px; text-align:left;'>{row[0]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[1]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[2]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[3]}</td>"; html_table += "</tr>"
        html_table += "</tbody></table><br>"; display(HTML(html_table))
    else: display(Markdown("*No ratio data available for table display.*"))

    # --- Display Ratio Comparison Chart (Plotly Bullet - V21.2 Fixes) ---
    if ratio_plotly_data['Ratio']:
       num_ratios = len(ratio_plotly_data['Ratio']); fig_ratios = go.Figure()
       good_color = '#90EE90'; bad_color = '#F08080'; avg_color = '#D3D3D3' # Lighter colors

       def get_range_colors(k):
           # Higher is better ratios
           if k in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio']: return [bad_color, avg_color, good_color] # Bad, Avg, Good
           # Valuation ratios (lower might be better vs benchmark)
           elif k in ['P/E','P/B','EV/EBITDA']: return [good_color, avg_color, bad_color] # Good, Avg, Bad (High multiple)
           else: return ['lightgrey','darkgrey','grey'] # Default grey scale

       for i in range(num_ratios):
           ratio_name = ratio_plotly_data['Ratio'][i];
           # Use the numeric values prepared for plotly (already in % form if applicable)
           company_val = ratio_plotly_data['Company'][i]; benchmark_val = ratio_plotly_data['Benchmark'][i];
           bench_src_plot = ratio_plotly_data['Benchmark Source'][i]
           is_percent = ratio_name in ['ROE', 'ROA', 'ROCE'] # Check if the current ratio needs % suffix

           # Skip if essential values are not finite numbers
           if not np.isfinite(company_val) or not np.isfinite(benchmark_val):
               print(f"Skipping plot for {ratio_name}: Invalid Value ({company_val}) or Bench ({benchmark_val}).")
               continue

           range_colors = get_range_colors(ratio_name);

           # Define step ranges using ADDITIVE approach (e.g., +/- 15 absolute points for %)
           range_delta = 15.0 if is_percent else max(1.0, abs(benchmark_val * 0.2)) # +/- 15% points, or 20% of value for others
           if ratio_name == 'Asset Turnover': range_delta = 0.2 # Specific delta for AT
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: range_delta = 0.5
           elif ratio_name in ['P/E', 'P/B', 'EV/EBITDA']: range_delta = max(2.0, abs(benchmark_val*0.2))

           range1_limit = benchmark_val - range_delta # Lower bound of 'average' range
           range2_limit = benchmark_val + range_delta # Upper bound of 'average' range

           # Determine plot axis range dynamically but capped
           abs_max_val = max(abs(company_val), abs(benchmark_val), 1) # Ensure > 0
           plot_max_val = max(company_val, benchmark_val) + range_delta * 1.5 # Extend beyond avg range
           plot_min_val = min(company_val, benchmark_val) - range_delta * 1.5

           # Apply sensible caps to ranges based on ratio type
           if ratio_name in ['P/E', 'EV/EBITDA']: plot_max_val = min(plot_max_val, 100); plot_min_val = max(plot_min_val, -20)
           elif ratio_name == 'P/B': plot_max_val = min(plot_max_val, 20); plot_min_val = max(plot_min_val, -5)
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: plot_max_val = min(plot_max_val, 5); plot_min_val = max(plot_min_val, 0)
           elif is_percent: plot_max_val = min(plot_max_val, 150); plot_min_val = max(plot_min_val,-100) # Wider % range
           elif ratio_name == 'Asset Turnover': plot_max_val = min(plot_max_val, 3); plot_min_val = max(plot_min_val, -0.5)

           # Ensure min < max
           if plot_min_val >= plot_max_val: plot_max_val = plot_min_val + range_delta # Add delta if equal

           # Define gauge steps based on benchmark and additive ranges
           steps = [];
           step_boundaries = sorted([plot_min_val, range1_limit, range2_limit, plot_max_val])
           step_colors = range_colors # Use the predefined color order

           # Create steps, ensuring start < end for each range
           for j in range(len(step_colors)):
               start_range = step_boundaries[j]; end_range = step_boundaries[j+1]
               if start_range < end_range: # Only add step if range is valid
                  steps.append({'range': [start_range, end_range], 'color': step_colors[j]})
               # else: print(f"Info: Invalid step range for {ratio_name} step {j}: [{start_range}, {end_range}]") # Debug

           # Handle benchmark near zero - make average range around zero
           if abs(benchmark_val) < 1e-9:
                range1_limit_zero = -range_delta / 2
                range2_limit_zero = range_delta / 2
                step_boundaries_zero = sorted([plot_min_val, range1_limit_zero, range2_limit_zero, plot_max_val])
                steps = []
                for j in range(len(step_colors)):
                    start_range = step_boundaries_zero[j]; end_range = step_boundaries_zero[j+1]
                    if start_range < end_range: steps.append({'range': [start_range, end_range], 'color': step_colors[j]})
                # print(f"Info: Benchmark for {ratio_name} is zero. Steps: {steps}") # Debug

           # Format benchmark for title (already in % form if needed)
           bench_fmt_title = f"{benchmark_val:.1f}%" if is_percent else f"{benchmark_val:.2f}";
           title_text = f"<b>{ratio_name}</b><br><i style='font-size:0.8em'>(Bench: {bench_fmt_title} [{bench_src_plot.replace('Fallback ','')}])</i>"

           # Define delta colors (using lighter palette)
           lower_is_better = ratio_name in ['P/E', 'P/B', 'EV/EBITDA']
           delta_decreasing_color = '#90EE90' if lower_is_better else '#F08080'; # Light Green / Light Red
           delta_increasing_color = '#F08080' if lower_is_better else '#90EE90'; # Light Red / Light Green

           # Add indicator trace
           fig_ratios.add_trace(go.Indicator(
                mode = "number+gauge+delta", value = company_val,
                number={'suffix': "%" if is_percent else "", 'valueformat': ".1f" if is_percent else ".2f", 'font':{'size':16}},
                # Delta shows absolute difference from benchmark
                delta = {'reference': benchmark_val, 'relative': False, 'valueformat':".1f" if is_percent else ".2f",
                         'decreasing': {'color': delta_decreasing_color}, 'increasing': {'color': delta_increasing_color}},
                domain = {'row': i, 'column': 0}, title = {'text' : title_text, 'font':{'size':11}},
                gauge = { 'shape': "bullet",
                         'axis': {'range': [plot_min_val, plot_max_val], 'tickformat':'.1f'}, # Consistent tick format
                         'threshold': { 'line': {'color': "black", 'width': 2}, 'thickness': 0.8, 'value': benchmark_val },
                         'steps': steps, # Use the generated steps list
                         'bar': {'color': "#2c3e50", 'thickness': 0.5} # Color of the company value bar
                       }
            ))
       fig_ratios.update_layout(
           height = max(300, num_ratios * 90), # Increased height per ratio slightly
           grid = {'rows': num_ratios, 'columns': 1, 'pattern': "independent"},
           title_text='Ratio Performance vs. Benchmarks', title_x=0.5,
           margin=dict(l=180, r=40, t=60, b=30) # Increased left margin for longer titles
           )
       fig_ratios.show()
    else: display(Markdown("*No valid ratios/benchmarks available for comparison chart.*"))

    # Return calculated ratios (still decimal form) and table data
    return ratios, ratio_data_for_table

# --- Execute Ratio Analysis ---
ratios_calculated = {}; ratio_table = []
# Ensure final_valuation is dict and doesn't have a fatal error preventing EV calculation
valuation_ready = isinstance(final_valuation, dict) and not (final_valuation.get('error') and "Cannot proceed" in final_valuation.get('error',''))
wacc_ready = isinstance(wacc_components, dict)

if valuation_ready and wacc_ready:
    try:
        ev_for_ratios = final_valuation.get('enterprise_value', np.nan) # Get calculated EV
        if not np.isfinite(ev_for_ratios): print("WARNING: EV for ratios is invalid. EV/EBITDA will fail.")
        ratios_calculated, ratio_table = analyze_ratios(
            stock_data['info'], stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'],
            ev_for_ratios, current_price, shares_outstanding, currency_symbol,
            user_benchmarks, benchmark_sources, wacc_components, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Ratio Analysis: {e}**"))
         print(traceback.format_exc())
         ratios_calculated = {'error': f'Runtime error in ratio analysis: {e}'}; ratio_table = []
else:
    error_reason = final_valuation.get('error', 'Unknown Valuation Error') if not valuation_ready else 'WACC Components Invalid'
    display(Markdown(f"❌ **Skipping Ratio Analysis due to prior errors:** {error_reason}"));
    ratios_calculated = {'error': f'Skipped due to: {error_reason}'}; ratio_table = []


# %% [markdown]
# ## 9. Summary & Visualization Dashboard
# (No changes needed in this section's logic based on the request, keeping previous robust version)

# %%
# Dashboard Function Definition
def display_summary_dashboard(company_nm, ticker_sym, current_pr, intrinsic_val_per_sh,
                              wacc_val, term_growth, init_growth, growth_src,
                              final_val_dict, hist_df, fcff_res, currency_sym, info_dict, input_src_dict,
                              is_indian): # Added flag

    display(Markdown(f"## Valuation Summary Dashboard: {company_nm} ({ticker_sym})"))
    # Build source summary string safely
    source_items = []
    for k, v in input_src_dict.items():
        if isinstance(v, str): source_items.append(f"{k}:`{v.split('(')[0].strip()}`")
    source_summary = " | ".join(source_items)

    growth_src_display = growth_src if isinstance(growth_src, str) else 'N/A'
    fcff_basis_method_disp = fcff_res.get('fcff_basis_method', 'N/A') if isinstance(fcff_res, dict) else 'N/A'
    display(Markdown(f"**Input Sources:** {source_summary} | Growth:`{growth_src_display}` | FCFF Basis:`{fcff_basis_method_disp}`"))

    # Check validity of key results
    valid_intrinsic = intrinsic_val_per_sh is not None and np.isfinite(intrinsic_val_per_sh)
    valid_current_pr = current_pr is not None and np.isfinite(current_pr) and current_pr > 0
    valid_wacc = wacc_val is not None and np.isfinite(wacc_val)
    valid_term_g = term_growth is not None and np.isfinite(term_growth)
    valid_init_g = init_growth is not None and np.isfinite(init_growth)

    # --- Indicator Subplots ---
    fig_indicators = make_subplots(rows=1, cols=3, specs=[[{'type': 'indicator'}, {'type': 'indicator'}, {'type': 'indicator'}]],
                                 subplot_titles=("Price vs. Intrinsic", "WACC", "Growth Rates"), horizontal_spacing=0.1, column_widths=[0.4, 0.3, 0.3])

    # Indicator 1: Price vs Intrinsic
    intrinsic_val_num = intrinsic_val_per_sh if valid_intrinsic else np.nan
    intrinsic_val_fmt = format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian) if valid_intrinsic else "N/A"
    current_pr_num = current_pr if valid_current_pr else 0.0; upside_fmt = ""; delta_config = None
    comparison_text = f"Intrinsic: {intrinsic_val_fmt}" # Start with base text

    if valid_intrinsic and valid_current_pr :
        if intrinsic_val_num > 1e-9: # Use small threshold instead of > 0 for comparison
             upside_pct = (intrinsic_val_num / current_pr_num) - 1
             upside_fmt = f" ({upside_pct:+.1%})"
             comparison_text += upside_fmt # Add upside % if intrinsic > 0
             # Delta shows difference FROM intrinsic value
             delta_config = {'reference': intrinsic_val_num, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym,
                             'decreasing': {'color': '#2ecc71', 'symbol': '▲ Upside'}, # Price < Intrinsic
                             'increasing': {'color': '#e74c3c', 'symbol': '▼ Downside'}} # Price > Intrinsic
        else: # Handle zero or negative intrinsic value
             comparison_text += " (≤ 0)" # Indicate non-positive intrinsic
             # Delta shows difference FROM zero
             delta_config = {'reference': 0, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym,
                             'decreasing': {'color': 'orange', 'symbol': ''}, # Price below zero (unlikely)
                             'increasing': {'color': '#e74c3c', 'symbol': '▼ Price > 0'}} # Price is above zero intrinsic
    elif valid_intrinsic: comparison_text += " (Current N/A)"
    else: comparison_text = f"Intrinsic: N/A (Calculation Error)" # Intrinsic value itself failed

    fig_indicators.add_trace(go.Indicator(mode = "number+delta" if delta_config and valid_current_pr else "number", # Only show delta if current price valid
        value = current_pr_num if valid_current_pr else 0,
        number={'prefix': currency_sym + " " if is_indian else currency_sym, 'valueformat': '.2f', 'font':{'size':24}},
        title = {"text": f"<span style='font-size:0.8em'>{comparison_text}</span>"},
        delta = delta_config if valid_current_pr else None, # Don't show delta if no current price
        domain = {'row': 0, 'column': 0}), row=1, col=1)

    # Indicator 2: WACC Gauge
    wacc_val_num = wacc_val * 100 if valid_wacc else None
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = wacc_val_num,
        number = {'suffix': "%", 'valueformat':'.1f' if valid_wacc else '', 'font':{'size':20}},
        gauge = {'axis': {'range': [MIN_WACC*100, MAX_WACC*100]}, 'bar': {'color': "darkblue" if valid_wacc else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [ {'range': [MIN_WACC*100, 9], 'color': "#d5f5e3"}, {'range': [9, 14], 'color': "#fdebd0"}, {'range': [14, MAX_WACC*100], 'color': "#fadbd8"}]} if valid_wacc else None,
        title = {"text": "<span style='font-size:0.8em'>WACC</span>"}, domain = {'row': 0, 'column': 1}), row=1, col=2)

    # Indicator 3: Growth Gauge
    term_growth_num = term_growth * 100 if valid_term_g else None; init_growth_fmt = format_percent(init_growth, 1) if valid_init_g else "N/A"
    growth_title_text = f"<span style='font-size:0.8em'>Terminal Growth (g)<br>(Initial: {init_growth_fmt})</span>"
    # Dynamic gauge range based on terminal growth
    gauge_max_range = max(TERMINAL_GROWTH_RATE_CAP * 100 + 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) + 2) # Ensure range goes beyond value
    gauge_min_range = min(MIN_GROWTH * 100 - 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) - 2) # Ensure range goes below value

    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = term_growth_num,
         number = {'suffix': "%", 'valueformat':'.1f' if valid_term_g else '', 'font':{'size':20}},
         gauge = {'axis': {'range': [gauge_min_range, gauge_max_range], 'tickwidth': 1}, 'bar': {'color': "darkgrey" if valid_term_g else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [{'range': [gauge_min_range, 2.0], 'color': "#fadbd8"}, {'range': [2.0, 4.5], 'color': "#fdebd0"}, {'range': [4.5, gauge_max_range], 'color': "#d5f5e3"}], # Lt Red, Lt Yellow, Lt Green
                 'threshold' : {'line': {'color': "black", 'width': 3}, 'thickness': 0.75, 'value': term_growth_num } if valid_term_g else None } if valid_term_g else None,
         title = {"text": growth_title_text}, domain = {'row': 0, 'column': 2}), row=1, col=3)

    fig_indicators.update_layout(height=220, margin=dict(l=30, r=30, t=50, b=20)); fig_indicators.show()

    # --- Historical Price Chart ---
    if hist_df is not None and not hist_df.empty:
        # Prefer 'Adj Close' if available and has data, else 'Close'
        price_col_hist = None
        if 'Adj Close' in hist_df.columns and hist_df['Adj Close'].notna().any(): price_col_hist = 'Adj Close'
        elif 'Close' in hist_df.columns and hist_df['Close'].notna().any(): price_col_hist = 'Close'

        if price_col_hist:
             fig_hist = px.line(hist_df, x=hist_df.index, y=price_col_hist, title=f'{company_nm} Price History ({price_col_hist}) vs. DCF');
             fig_hist.update_layout(xaxis_title='Date', yaxis_title=f'Price ({currency_sym})', height=400, legend_title_text='Legend'); fig_hist.update_xaxes(rangeslider_visible=False)
             # Add lines only if values are valid
             if valid_intrinsic:
                 fig_hist.add_hline(y=intrinsic_val_num, line_dash="dash", line_color="#e74c3c", opacity=0.8, annotation_text=f"DCF Intrinsic: {format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian)}", annotation_position="bottom right", annotation_font_color="#e74c3c")
             if valid_current_pr:
                 fig_hist.add_hline(y=current_pr_num, line_dash="dot", line_color="#3498db", opacity=0.8, annotation_text=f"Current: {format_currency(current_pr_num, currency_sym, is_indian=is_indian)}", annotation_position="top left", annotation_font_color="#3498db")
             fig_hist.show()
        else: display(Markdown("*Historical price data ('Adj Close' or 'Close') unavailable or invalid for chart.*"))
    else: display(Markdown("*Historical price data object unavailable.*"))

    # --- DCF EV Breakdown Pie Chart ---
    if isinstance(fcff_res, dict):
        pv_forecast_sum = fcff_res.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_res.get('pv_terminal_value', np.nan)
        # Check if components are finite numbers
        valid_pvs = np.isfinite([pv_forecast_sum, pv_tv]);
        total_ev = np.nansum([pv_forecast_sum, pv_tv]) if any(valid_pvs) else np.nan # Sum if at least one is valid

        # Proceed if EV is a finite number and there wasn't a fatal FCFF error
        fcff_fatal_error = fcff_res.get('error') and "Cannot proceed" in fcff_res.get('error', '')
        if np.isfinite(total_ev) and not fcff_fatal_error:
            labels_pie = []; values_pie = []; pull_pie = []; colors = ['#4e79a7','#f28e2b'] # Blue, Orange
            # Only include components > 0 to avoid tiny negative slices
            if abs(pv_forecast_sum) > 1e-6 : labels_pie.append(f'PV({FORECAST_YEARS}yr FCFFs)'); values_pie.append(abs(pv_forecast_sum)); pull_pie.append(0) # Plot absolute value
            if abs(pv_tv) > 1e-6 : labels_pie.append('PV(TV)'); values_pie.append(abs(pv_tv)); pull_pie.append(0.05 if len(labels_pie) > 1 else 0) # Plot absolute value

            if labels_pie: # Check if there's anything to plot
                hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                if is_indian: hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>" # No change needed here
                fig_dcf_pie = go.Figure(data=[go.Pie(labels=labels_pie, values=values_pie, hole=.4, pull=pull_pie, marker_colors=colors[:len(labels_pie)], hovertemplate=hover_format, textinfo='percent+label', insidetextorientation='radial', sort=False)])
                fig_dcf_pie.update_layout( title_text=f"DCF EV Breakdown (Total EV: {format_currency(total_ev, currency_sym, is_indian=is_indian)})", title_x=0.5, legend_title_text='Components (Absolute Value)', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), height=350, margin=dict(t=60, b=50, l=20, r=20) )
                fig_dcf_pie.show()
            else: display(Markdown(f"*DCF components zero/negligible or negative. Cannot generate EV Breakdown chart.*"))
        elif fcff_fatal_error:
             display(Markdown(f"*Cannot generate EV Breakdown chart due to fatal FCFF projection error: {fcff_res.get('error')}*"))
        else:
             display(Markdown(f"*EV calculation failed or invalid ({format_currency(total_ev, currency_sym, is_indian=is_indian)}). Cannot generate EV Breakdown chart.*"))
    else: display(Markdown(f"*FCFF results invalid or unavailable. Cannot generate EV Breakdown chart.*"))

    # --- FCFF Projection Bar Chart ---
    if isinstance(fcff_res, dict) and fcff_res.get('projected_fcffs') and not fcff_fatal_error:
        years = [f'Y{i+1}' for i in range(len(fcff_res['projected_fcffs']))];
        # Clean data, replacing non-finite with 0 for plotting
        pv_list_clean = [p if np.isfinite(p) else 0 for p in fcff_res.get('pv_fcffs', [])];
        fcff_list_clean = [f if np.isfinite(f) else 0 for f in fcff_res.get('projected_fcffs', [])]

        if len(pv_list_clean) == len(years) and len(fcff_list_clean) == len(years) and len(years) > 0:
            df_fcff_plot = pd.DataFrame({'Year': years, 'Forecasted FCFF': fcff_list_clean, 'PV of FCFF': pv_list_clean});
            # Ensure numeric types after cleaning
            df_fcff_plot['Forecasted FCFF'] = pd.to_numeric(df_fcff_plot['Forecasted FCFF']);
            df_fcff_plot['PV of FCFF'] = pd.to_numeric(df_fcff_plot['PV of FCFF'])
            # Melt dataframe for easy plotting with Plotly Express
            df_fcff_melt = df_fcff_plot.melt(id_vars='Year', var_name='Cash Flow Type', value_name='Value')
            fig_fcff_bar = px.bar(df_fcff_melt, x='Year', y='Value', color='Cash Flow Type', title='FCFF Forecast & Present Values', labels={'Value':f'Value ({currency_sym})'}, barmode='group', height=350, color_discrete_map={'Forecasted FCFF':'#5dade2', 'PV of FCFF':'#af7ac5'}) # Blue, Purple

            # Customize hover text
            hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            if is_indian: hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            fig_fcff_bar.update_traces(hovertemplate=hover_bar_format); fig_fcff_bar.update_layout(margin=dict(t=60, b=30, l=50, r=30)); fig_fcff_bar.show()
        else: display(Markdown("*FCFF projection data mismatch/invalid or empty. Cannot plot.*"))
    elif fcff_fatal_error:
         display(Markdown(f"*Cannot generate FCFF Projection chart due to fatal error: {fcff_res.get('error')}*"))
    else: display(Markdown("*FCFF projection data unavailable or invalid.*"))

    # --- Valuation Conclusion ---
    display(Markdown("### Valuation Conclusion"));
    if isinstance(final_val_dict, dict):
        intrinsic_val = final_val_dict.get('intrinsic_value_per_share', np.nan)
        val_error = final_val_dict.get('error', None) # Check for valuation specific errors
        fcff_error = fcff_res.get('error', None) if isinstance(fcff_res, dict) else "FCFF Data Invalid" # Check FCFF errors
        conclusion = ""

        if val_error: conclusion = f"<span style='color:red'>❌ Valuation Error: {val_error}</span>"
        elif fcff_fatal_error: conclusion = f"<span style='color:red'>❌ Valuation Failed: {fcff_error}</span>" # If FCFF stopped projection
        elif not valid_intrinsic: conclusion = f"<span style='color:red'>❌ Valuation Error: Intrinsic Value calculation failed (result is {intrinsic_val}).</span>"
        elif valid_intrinsic and intrinsic_val <= 0:
             conclusion = f"<span style='color:orange'>⚠️ Negative/Zero DCF Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}.</span> Suggests negative projected cash flows or high debt/negative EV. Standard DCF may be inappropriate."
        elif valid_intrinsic and not valid_current_pr:
             conclusion = f"✅ DCF Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}. Cannot compare (Current Price N/A)."
        elif valid_intrinsic and valid_current_pr : # Standard comparison
             diff_pct = (intrinsic_val / current_pr_num) - 1
             current_fmt = format_currency(current_pr_num, currency_sym, is_indian=is_indian)
             intrinsic_fmt = format_currency(intrinsic_val, currency_sym, is_indian=is_indian)
             if abs(diff_pct) < 0.10: conclusion = f"<span style='color:grey'>~ Fairly Valued:</span> Price ({current_fmt}) vs DCF ({intrinsic_fmt}). Diff: {diff_pct:+.1%}"
             elif diff_pct < 0: # Intrinsic < Price
                  overval_pct = (current_pr_num / intrinsic_val) - 1 if intrinsic_val != 0 else float('inf')
                  conclusion = f"<span style='color:#e74c3c'>🔻 Potentially Overvalued:</span> Price ({current_fmt}) is **{overval_pct:+.1%}** above DCF ({intrinsic_fmt})."
             else: # Intrinsic > Price
                  conclusion = f"<span style='color:#2ecc71'>✅ Potentially Undervalued:</span> Price ({current_fmt}) suggests **{diff_pct:+.1%}** upside to DCF ({intrinsic_fmt})."
        else: # Fallback if something unexpected happened
             conclusion = f"<span style='color:red'>❌ Valuation Error: Unexpected state (Intrinsic: {intrinsic_val}, Price: {current_pr_num}).</span>"
    else:
        conclusion = f"<span style='color:red'>❌ Valuation Error: Final valuation data structure invalid.</span>"
    display(Markdown(f"> **{conclusion}**"))

# --- Execute Summary Dashboard Generation ---
# Check if final_valuation exists, is dict, and does NOT have a fatal error
valuation_successful = ( isinstance(final_valuation, dict) and
                        not (final_valuation.get('error') and "Cannot proceed" in final_valuation.get('error','')) and
                        not (final_valuation.get('error') and "prevents valuation" in final_valuation.get('error','')) )


if valuation_successful and isinstance(fcff_results, dict) and isinstance(stock_data.get('info'), dict) and isinstance(input_sources, dict):
    try:
        display_summary_dashboard(
            company_name, ticker_symbol, current_price, final_valuation.get('intrinsic_value_per_share'),
            discount_rate, fcff_results.get('terminal_growth_rate'), fcff_results.get('initial_growth_rate'),
            fcff_results.get('growth_source', 'N/A'),
            final_valuation, # Pass the whole dict
            stock_data['history'], # Use 2y history for plot
            fcff_results, # Pass the whole dict
            currency_symbol, stock_data['info'], input_sources, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Dashboard Generation: {e}**"))
         print(traceback.format_exc()) # Print detailed traceback
else:
    # Provide specific reason for skipping
    skip_reason = final_valuation.get('error', 'Unknown upstream error') if isinstance(final_valuation, dict) else "Final valuation data invalid"
    if isinstance(fcff_results,dict) and fcff_results.get('error') and "Cannot proceed" in fcff_results.get('error'):
        skip_reason = fcff_results.get('error') # Prioritize fatal FCFF error message

    display(Markdown(f"❌ **Skipping Summary Dashboard due to prior errors:** {skip_reason}"));


# %% [markdown]
# ## 10. Assumptions & Disclaimer (Reflects V21.2 Logic)

# %%
# Dynamically generate assumption strings based on final values
rf_source_disp = input_sources.get('Rf', '?')
rm_source_disp = input_sources.get('Rm','?')
beta_source_disp = input_sources.get('Beta','?')
c_spread_source_disp = input_sources.get('Credit Spread','N/A')
bench_source_disp = input_sources.get('Benchmarks','?')

# Safely get values from potentially missing dictionaries
growth_source_disp = fcff_results.get('growth_source', 'N/A') if isinstance(fcff_results, dict) else 'Error/Not Run'
tax_rate_source_disp = wacc_components.get('tax_rate_source', 'N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
equity_source_disp = wacc_components.get('equity_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
weights_source_disp = wacc_components.get('weights_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
terminal_growth_final = fcff_results.get('terminal_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan
initial_growth_final = fcff_results.get('initial_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan
fcff_basis_method_disp = fcff_results.get('fcff_basis_method', 'N/A') if isinstance(fcff_results, dict) else 'Error/Not Run'


# %%
# ----> UPDATED TEXT TO REFLECT NEW FCFF BASIS LOGIC (V21.2) <----
display(Markdown(f"""
**Key Assumptions & Input Sources:**
*   **Financial Data:** Relies on data provided by Yahoo Finance (`yfinance`). Accuracy subject to provider. Missing data uses fallbacks or causes errors. Assumes annual statements unless only quarterly are available.
*   **Risk-Free Rate (Rf):** Attempted dynamic fetch, **confirmed/overridden by user.** (`Source: {rf_source_disp}` - Default suggestion: `{DEFAULT_RF_RATE_SUGGESTION:.1f}%`)
*   **Expected Market Return (Rm):** Based on **user input.** (`Source: {rm_source_disp}`)
*   **Market Risk Premium (MRP):** Derived from user `Rm` and final `Rf`.
*   **Beta (β):** Dynamically fetched if valid, otherwise **required user input** (defaults to 1.0 if fetch/input fails). (`Source: {beta_source_disp}`)
*   **Cost of Debt (Kd):** Implied from Interest Exp/Total Debt if possible & within bounds (`{format_percent(MIN_COST_OF_DEBT)}` - `{format_percent(MAX_COST_OF_DEBT)}`). Otherwise, defaults to `Rf + Credit Spread` (bounded).
*   **Credit Spread:** Used only if default Kd triggered. Based on **user input** (default suggestion: `{format_percent(DEFAULT_CREDIT_SPREAD)}`). (`Source: {c_spread_source_disp}`)
*   **FCFF Calculation (T-0 Basis for Projection):**
    1.  Calculates potential 4-year average FCFF using multiple methods (EBITDA-Tax-WC-Capex, NOPAT, CFO+IntAdj, CFO-Capex, Avg(EBITDA-TaxProv), Avg(EBITDAMargin)*RevT0).
    2.  If one or more methods yield a **positive average**, these options are **presented to the user for selection.**
    3.  If no averaging method yields a positive result, the T-0 individual best estimate (NOPAT -> CFO+IntAdj -> CFO-Capex) is used as a fallback.
    4.  **Projection halts if no method (including T-0 fallback) yields a positive basis.**
    *   Method Used: `{fcff_basis_method_disp}`.
    *   Historical FCFFs *for CAGR* use standard single-year fallbacks (NOPAT -> CFO+ -> CFO-).
*   **Growth Rates:**
    *   **Initial Growth:** `{format_percent(initial_growth_final)}` used for Year 1, determined *independently* of the T-0 projection basis: (`Source: {growth_source_disp}`)
        1.  Fetch last 4 data points (T-0 to T-3) of Revenue, FCFF (individual best estimate), and Stock Price (daily 'Adj Close' or 'Close').
        2.  Calculate **3-Year CAGR** (T-0 vs T-3) *only* for metrics with valid, positive start data (`T-3 > 0`).
        3.  Calculate the **Geometric Mean** (`geometric_mean_modified`) of successfully calculated, non-extreme CAGRs. Excludes CAGRs <-90% or >500%.
        4.  If GM calculation fails or *no valid/reasonable CAGRs were computed*, use **Default (`{format_percent(DEFAULT_GROWTH_RATE)}`)**.
        5.  Apply **Bounds:** The result is capped between **`{format_percent(MIN_INITIAL_GROWTH_FLOOR)}` (Floor)** and **`{format_percent(MAX_INITIAL_GROWTH_CEILING)}` (Ceiling)**.
    *   **Terminal Growth (g):** `{format_percent(terminal_growth_final)}` = min(GDP cap=`{format_percent(TERMINAL_GROWTH_RATE_CAP)}`, WACC*`0.60`, Final Initial Growth, Rf*`0.90`), floored at `{format_percent(MIN_GROWTH)}`, ensured `< WACC`. Calculated only if projection proceeds.
*   **Tax Rate:** Effective rate derived from latest financials ({tax_rate_source_disp}) (bounded `{format_percent(MIN_TAX_RATE)}` - `{format_percent(MAX_TAX_RATE)}`) or default (`{format_percent(DEFAULT_TAX_RATE)}`).
*   **Capital Structure:** Assumed stable using `{equity_source_disp}` and calculated debt. (`Weights Source: {weights_source_disp}`)
*   **Industry Benchmarks:** Uses **user input** (e.g., '15' for 15%) or **generic fallbacks** if input left blank. (`Source: {bench_source_disp}`)
*   **Currency Formatting:** Uses ₹/Lakh/Crore for `.NS`/`.BO` tickers, standard K/M/B/T otherwise.

**Limitations:**
*   **Model Sensitivity:** DCF highly sensitive to inputs (WACC, growth rates). **Sensitivity analysis (not included) crucial.**
*   **Input Quality:** Validity depends on user inputs and appropriateness of fallbacks. Benchmarks should be carefully chosen.
*   **Historical Data Availability:** Financial statement data via `yfinance` often limited (~4 annual points). Accuracy of 3-year CAGR depends heavily on these few points. CAGR calculation skipped if start/end data invalid or start non-positive.
*   **FCFF Calculation & Projection:** FCFF calculations rely on available data keys and can vary significantly between methods. The chosen projection basis depends on averaging methods yielding positive results and user selection. If financials are poor (consistent negative FCFF), the model **cannot project FCFF** and valuation fails. Averaging smooths results.
*   **CAGR/GM Modifications:** Modified CAGR requires positive start value. Modified GM handles negative rates but falls back to arithmetic mean on errors or extreme inputs. GM is less meaningful with highly disparate rates.
*   **Dynamic Data Reliability:** Depends on `yfinance` accuracy/availability. Rf proxies need verification. Stock history column names ('Adj Close', 'Close') assumed.
*   **Simplifications:** Linear growth decay, constant tax rate/structure in projection. Ignores qualitative factors, macro shocks, share buybacks/issuance during projection.
*   **Fallback Accuracy:** Default growth rate (`{format_percent(DEFAULT_GROWTH_RATE)}`) and generic benchmarks may not be representative.
*   **Data Completeness:** Performance degrades with missing data. *Cross-verify key data (e.g., Shares Outstanding, Debt).*

**Disclaimer:**
*For **educational and illustrative purposes only**. **NOT investment advice**. Models have limitations and rely on potentially imperfect data and assumptions. Users MUST conduct own due diligence and consult a qualified financial advisor before investing. Creators NOT liable for actions based on this report.*
"""))



# %%


````

````python
# %% [markdown]
# # Dynamic Stock Financial Analysis & DCF Valuation
# **(Interactive Report - Dynamic/User Input Driven Assumptions - V21.3 - User Method Choice / 3yr Projection)**
#
# **Analyst:** Simulated AI Research Analyst
#
# **Objective:** To perform a comprehensive financial analysis and estimate the intrinsic value per share using DCF. **User selects a single FCFF calculation method upfront.** Historical FCFF, average basis, and FCFF CAGR are calculated *only* using this selected method. Projection halts if the 4yr average basis is not positive. **Projection horizon set to 3 years.** Initial growth uses GM of Rev CAGR, Stock CAGR, and FCFF CAGR (from selected method).
#
# **Methodology:**
# 1.  **User Input & Dynamic Fetching:** Standard setup.
# 2.  **Data Acquisition:** Standard setup, including 4y daily history.
# 3.  **WACC Calculation:** Standard setup.
# 4.  **User FCFF Method Selection:** Prompt user to choose one primary FCFF calculation method (0-4).
# 5.  **FCFF Calculation & Projection:**
#     *   **Calculate Historical FCFF (T-0 to T-3):** Use new helper `calculate_fcff_single_method_single_year` based *only* on user's choice.
#     *   **Determine Projection Basis:** Calculate the 4-year average of the historical FCFFs (from selected method).
#     *   **Halt if Basis Not Positive:** If the 4yr average is <= 0, set error and stop FCFF process.
#     *   **Estimate initial growth:** Calculate 3yr CAGRs for Rev, Stock, and FCFF (using selected method's history). Calculate GM. Fallback/Bound.
#     *   **Project FCFF (3 Years):** Project forward from the positive average basis for **3 years** using decaying growth rate.
# 6.  **Terminal Value:** Standard calculation based on Year 3 FCFF.
# 7.  **Discounting & Valuation:** Standard calculation.
# 8.  **Ratio Analysis:** Standard calculation (Plotting logic corrected).
# 9.  **Interactive Visualization:** Standard setup.

# %% [markdown]
# ## 1. Setup and Libraries

# %%
# Standard Libraries
import warnings
import sys
import numpy as np
import pandas as pd
import datetime
import math # For isnan, isinf checks
import traceback # For detailed error logging

# Data Fetching
import yfinance as yf

# Visualization (Using Plotly)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

# Display Utilities
from IPython.display import display, Markdown, HTML

# Settings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pio.templates.default = "plotly_white" # Cleaner Plotly template

# %% [markdown]
# ## 2. Configuration and Helper Functions (FCFF Helpers Updated V21.3)

# %%
# --- Model Configuration ---
DEFAULT_TICKER = "MUTHOOTFIN.NS"
FORECAST_YEARS = 3 # MODIFIED: Project only 3 years
TERMINAL_GROWTH_RATE_CAP = 0.05
DEFAULT_GROWTH_RATE = 0.065 # Used if CAGR/GM fails
DEFAULT_RF_RATE_SUGGESTION = 7.0
DEFAULT_CURRENCY_SYMBOL = "₹"
DEFAULT_TAX_RATE = 0.25
DEFAULT_CREDIT_SPREAD = 0.015 # 1.5%
DEFAULT_MARKET_RISK_PREMIUM = 0.06 # 6.0%

# --- Generic Fallback Benchmark Ratios ---
DEFAULT_BENCHMARK_RATIOS = {
    'P/E': 17.3, 'P/B': 2.5, 'EV/EBITDA': 11.0, 'ROE': 19.05,
    'ROA': 3.92, 'ROCE': 11.85,'Asset Turnover': 0.15,
    'Current Ratio': 1.5, 'Quick Ratio': 1.5
}

# --- Model Stability Bounds ---
MIN_WACC = 0.06; MAX_WACC = 0.25; MIN_TAX_RATE = 0.10; MAX_TAX_RATE = 0.40
MIN_GROWTH = 0.00; MIN_INITIAL_GROWTH_FLOOR = 0.011; MAX_INITIAL_GROWTH_CEILING = 0.15
MIN_COST_OF_DEBT = 0.03; MAX_COST_OF_DEBT = 0.18

# --- Ratio Keys Needed for Benchmarking ---
REQUIRED_BENCHMARK_RATIOS = list(DEFAULT_BENCHMARK_RATIOS.keys())

# --- FCFF Method Definitions (for User Selection) ---
FCFF_METHODS_DEFINITION = {
    0: {"name": "EBITDA - TaxProv - WC - Capex", "formula": "EBITDA - Tax Provision - Change in WC - CapEx"},
    1: {"name": "NOPAT + Depr - WC - Capex", "formula": "EBIT*(1-Tax Rate) + Depr&Amort - Change in WC - CapEx"},
    2: {"name": "CFO + IntExp*(1-T) - Capex", "formula": "Cash Flow Ops + Interest Exp*(1-Tax Rate) - CapEx"},
    3: {"name": "CFO - Capex", "formula": "Cash Flow Ops - CapEx"},
    4: {"name": "EBITDA - TaxProv (Proxy)", "formula": "EBITDA - Tax Provision (Simplified Proxy - Ignores WC/Capex)"}
}


# --- Helper Functions --- (Keep relevant ones, remove unused FCFF fallbacks)
def print_keys(statement, name):
    # (Unchanged)
    print(f"\n--- Available Keys in {name} ---")
    if isinstance(statement, pd.DataFrame):
        valid_indices = [idx for idx in statement.index if idx is not None]
        print(", ".join(map(str, valid_indices)))
    elif isinstance(statement, dict): print(", ".join(statement.keys()))
    print("------------------------------------\n")

def safe_get(data, key_list, statement_name="Statement", default=np.nan):
    # (Unchanged - Robust fetching is still useful)
    if not isinstance(key_list, list): key_list = [key_list]
    value = None
    found_key = None
    for key in key_list:
        current_value = None; raw_val = None
        data_source = None
        if isinstance(data, pd.Series):
            data_source = f"'{statement_name}' Series"
            if key in data.index:
                try:
                    raw_val = data.loc[key]
                    if isinstance(raw_val, (int, float)) and np.isfinite(raw_val): current_value = float(raw_val)
                    elif isinstance(raw_val, pd.Series):
                        if not raw_val.empty and pd.notna(raw_val.iloc[0]) and np.isfinite(raw_val.iloc[0]): current_value = float(raw_val.iloc[0])
                    elif pd.notna(raw_val):
                        try_val = float(raw_val);
                        if np.isfinite(try_val): current_value = try_val
                except (KeyError, TypeError, IndexError, ValueError): pass
        elif isinstance(data, dict):
            data_source = f"'{statement_name}' Dict"
            raw_val = data.get(key)
            if raw_val is not None:
                try:
                    if isinstance(raw_val, str) and raw_val.lower() in ['n/a', 'none', '', '--', '-']: pass
                    else: temp_val = float(raw_val);
                    if np.isfinite(temp_val): current_value = temp_val
                except (TypeError, ValueError): pass
        if current_value is not None:
            value = current_value;
            found_key = key
            break
    if value is None: value = default
    return value

def format_currency(value, symbol="₹", precision=2, is_indian=False):
    # (Unchanged)
    if value is None or pd.isna(value) or not math.isfinite(value): return "N/A"
    if abs(value) > 1e18: return "Inf"
    sign = "-" if value < 0 else ""; abs_val = abs(value)
    if is_indian:
        symbol = "₹"
        if abs_val >= 1e7: formatted_val = f"{abs_val / 1e7:.{precision}f} Cr"
        elif abs_val >= 1e5: formatted_val = f"{abs_val / 1e5:.{precision}f} Lakh"
        else: formatted_val = f"{abs_val:,.{precision}f}"
        return f"{sign}{symbol} {formatted_val}"
    else:
        if abs_val >= 1e12: formatted_val = f"{abs_val / 1e12:.{precision}f}T"
        elif abs_val >= 1e9: formatted_val = f"{abs_val / 1e9:.{precision}f}B"
        elif abs_val >= 1e6: formatted_val = f"{abs_val / 1e6:.{precision}f}M"
        elif abs_val >= 1e3: formatted_val = f"{abs_val / 1e3:.{precision}f}K"
        else: formatted_val = f"{abs_val:.{precision}f}"
        return f"{sign}{symbol}{formatted_val}"

def format_percent(value, precision=1):
    # (Unchanged)
    if value is None or pd.isna(value) or not math.isfinite(value): return "N/A"
    if abs(value) > 1e5: return "Inf"
    return f"{value * 100:.{precision}f}%"

def get_validated_float_input(prompt_message, default_value=None, allow_blank=False, min_val=None, max_val=None, is_percentage=False):
    # (Unchanged)
    while True:
        prompt_with_default = prompt_message
        if is_percentage: prompt_with_default += " (e.g., enter '15' for 15%)"
        if default_value is not None: prompt_with_default += f" [Default: {default_value}]: "
        elif allow_blank: prompt_with_default += f" [Leave blank to skip/use fallback]: "
        else: prompt_with_default += ": "
        try:
            user_input_str = input(prompt_with_default).strip()
            if not user_input_str:
                if allow_blank: return None
                if default_value is not None:
                    try: return float(default_value)
                    except ValueError: raise ValueError(f"Internal Error: Invalid default value '{default_value}'.")
                raise ValueError("Input cannot be empty.")
            if user_input_str.endswith('%'): user_input_str = user_input_str[:-1]
            value = float(user_input_str)
            if min_val is not None and value < min_val: raise ValueError(f"Value must be >= {min_val}.")
            if max_val is not None and value > max_val: raise ValueError(f"Value must be <= {max_val}.")
            return value
        except ValueError as e: print(f"❌ Invalid input: {e}. Please enter a valid number.")

def try_get_dynamic_rf(rf_ticker_symbol="^TNX"):
    # (Unchanged)
    print(f"Attempting to fetch dynamic Risk-Free Rate ({rf_ticker_symbol})...")
    try:
        rf_ticker = yf.Ticker(rf_ticker_symbol); hist = rf_ticker.history(period="5d")
        if not hist.empty and 'Close' in hist.columns:
            last_close = hist['Close'].iloc[-1]
            if pd.notna(last_close):
                dynamic_rf = last_close / 100.0
                if 0.001 < dynamic_rf < 0.20: print(f"✅ Dynamically fetched Rf ({rf_ticker_symbol}): {format_percent(dynamic_rf)}"); return dynamic_rf
                else: print(f"⚠️ Fetched Rf value ({format_percent(dynamic_rf)}) out of typical range (0.1%-20%). Ignoring."); return None
            else: print(f"⚠️ Fetched Rf data was invalid (NaN) for {rf_ticker_symbol}."); return None
        else: print(f"⚠️ Could not fetch recent data for {rf_ticker_symbol}."); return None
    except Exception as e: print(f"❌ Error fetching dynamic Rf ({rf_ticker_symbol}): {e}"); return None

# --- Growth Calculation Helpers (Unchanged) ---
def calculate_cagr_modified(start_val, end_val, num_years):
    # (Unchanged)
    if pd.isna(start_val) or pd.isna(end_val) or not math.isfinite(start_val) or not math.isfinite(end_val) or num_years <= 0: return np.nan
    if start_val <= 1e-9: return np.nan
    cagr = np.nan
    try:
        if end_val <= 1e-9:
            end_val_abs = max(abs(end_val), 1e-9); start_val_abs = abs(start_val)
            ratio = end_val_abs / start_val_abs
            if ratio > 0: cagr_magnitude = ratio ** (1 / num_years) - 1; cagr = -cagr_magnitude
        else:
            ratio = end_val / start_val
            if ratio > 0: cagr = ratio ** (1 / num_years) - 1
    except (ValueError, OverflowError, ZeroDivisionError) as e: cagr = np.nan
    if pd.notna(cagr) and not (-0.999 < cagr < 20.0): cagr = np.nan
    return cagr

def geometric_mean_modified(rates):
    # (Unchanged)
    valid_rates = [r for r in rates if pd.notna(r) and math.isfinite(r)]
    if not valid_rates: return np.nan
    if len(valid_rates) == 1: return valid_rates[0]
    offset = 1.0; factors_offset = [r + offset for r in valid_rates]
    if any(f <= 1e-9 for f in factors_offset):
         print("⚠️ Geometric mean calculation failed due to rate <= -100%. Falling back to arithmetic mean.")
         return np.mean(valid_rates)
    n = len(valid_rates); product = 1.0
    try:
        for factor in factors_offset: product *= factor
        if product < 0 and n % 2 == 0:
             print("⚠️ Geometric mean calculation failed (negative base for even root). Falling back to arithmetic mean.")
             return np.mean(valid_rates)
        gm_offset = np.sign(product) * (np.abs(product) ** (1/n)); gm = gm_offset - offset
        if pd.notna(gm) and not (-0.999 < gm < 20.0):
            print(f"⚠️ Geometric mean result ({format_percent(gm)}) out of reasonable bounds. Falling back to arithmetic mean.")
            return np.mean(valid_rates)
        return gm
    except (ValueError, OverflowError, ZeroDivisionError, TypeError) as e:
        print(f"⚠️ Geometric mean calculation encountered numeric error ({e}). Falling back to arithmetic mean.")
        return np.mean(valid_rates)

# --- FCFF Component Fetcher (Still useful for grabbing raw data) ---
def _get_fcff_components(fin_col, bs_col, cf_col, bs_prev_col, log_list):
    # (Mostly Unchanged, logging kept from previous version)
    components = {}
    log_list.append("--- Fetching FCFF Components ---")

    # Revenue
    rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
    components['revenue'] = safe_get(fin_col, rev_keys, "Income Stmt", default=np.nan)
    if np.isfinite(components['revenue']): log_list.append(f"    Revenue: {format_currency(components['revenue'], '', 0)}")
    else: log_list.append(f"    Revenue: N/A (Tried: {', '.join(rev_keys)})")

    # EBIT
    ebit_keys = ["EBIT", "Ebit", "Operating Income", "Total Operating Income As Reported", "Earnings Before Interest and Tax"]
    components['ebit'] = safe_get(fin_col, ebit_keys, "Income Stmt", default=np.nan)
    if np.isfinite(components['ebit']): log_list.append(f"    EBIT: {format_currency(components['ebit'], '', 0)}")
    else:
        # Fallback: EBIT = Pretax Income + Interest Expense
        pretax_inc_keys = ['Pretax Income', 'EBT', 'Earnings Before Tax']
        pretax_inc = safe_get(fin_col, pretax_inc_keys, "Income Stmt (EBIT Fallback)", default=np.nan)
        interest_exp_keys = ['Interest Expense', 'Interest Expense Net']
        int_exp_for_ebit = abs(safe_get(fin_col, interest_exp_keys, "Income Stmt (EBIT Fallback)", default=np.nan))
        if np.isfinite(pretax_inc) and np.isfinite(int_exp_for_ebit):
            components['ebit'] = pretax_inc + int_exp_for_ebit
            log_list.append(f"    EBIT (Derived): {format_currency(pretax_inc,'',0)} + {format_currency(int_exp_for_ebit,'',0)} = {format_currency(components['ebit'],'',0)}")
        else:
            log_list.append(f"    EBIT: N/A (Primary & Fallback Failed)")
            components['ebit'] = np.nan

    # Depreciation & Amortization
    dep_amort_keys = ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"]
    components['dep_amort'] = safe_get(cf_col, dep_amort_keys, "Cash Flow", default=np.nan)
    if not np.isfinite(components['dep_amort']):
         components['dep_amort'] = safe_get(fin_col, ["Depreciation And Amortization", "Depreciation Income Statement"], "Income Stmt (Dep Fallback)", default=np.nan)
         if np.isfinite(components['dep_amort']): log_list.append(f"    Dep&Amort (IS Fallback): {format_currency(components['dep_amort'], '', 0)}")
         else: log_list.append(f"    Dep&Amort: N/A"); components['dep_amort'] = np.nan # Ensure NaN if not found
    else: log_list.append(f"    Dep&Amort (CF): {format_currency(components['dep_amort'], '', 0)}")

    # EBITDA (Calculated)
    if np.isfinite(components.get('ebit')) and np.isfinite(components.get('dep_amort')):
        components['ebitda'] = components['ebit'] + components['dep_amort']
        log_list.append(f"    EBITDA (Calc): {format_currency(components['ebitda'],'',0)}")
    else: components['ebitda'] = np.nan; log_list.append(f"    EBITDA: N/A")

    # CapEx
    capex_keys = ["Capital Expenditure", "Net Capital Expenditure", "Purchase Of Property Plant Equipment", "Net PPE Purchase And Sale", "Capital Expenditures"]
    capex_val = safe_get(cf_col, capex_keys, "Cash Flow", default=np.nan)
    components['capex'] = abs(capex_val) if np.isfinite(capex_val) else np.nan
    if np.isfinite(components['capex']): log_list.append(f"    CapEx (Abs): {format_currency(components['capex'],'',0)} (Raw: {format_currency(capex_val,'',0)})")
    else: log_list.append(f"    CapEx: N/A")

    # Change in Working Capital (WC)
    wc_keys = ["Change In Working Capital", "Change In Operating Capital", "Change In Operating Assets And Liabilities", "Change In Receivables"]
    components['wc_cf'] = safe_get(cf_col, wc_keys, "Cash Flow (Direct WC)", default=np.nan)
    if np.isfinite(components['wc_cf']):
        components['wc'] = components['wc_cf']; components['wc_src'] = "CF Direct"
        log_list.append(f"    Change in WC (CF Direct): {format_currency(components['wc'], '', 0)}")
    else:
        # Fallback to BS calculation
        components['wc_src'] = "BS Calc"; components['wc'] = np.nan
        if isinstance(bs_prev_col, pd.Series) and isinstance(bs_col, pd.Series):
             curr_assets = safe_get(bs_col, ['Current Assets', 'Total Current Assets'], 'BS Curr', default=np.nan)
             curr_liab = safe_get(bs_col, ['Current Liabilities', 'Total Current Liabilities'], 'BS Curr', default=np.nan)
             prev_assets = safe_get(bs_prev_col, ['Current Assets', 'Total Current Assets'], 'BS Prev', default=np.nan)
             prev_liab = safe_get(bs_prev_col, ['Current Liabilities', 'Total Current Liabilities'], 'BS Prev', default=np.nan)
             if np.isfinite(curr_assets) and np.isfinite(curr_liab) and np.isfinite(prev_assets) and np.isfinite(prev_liab):
                 curr_wc_val = curr_assets - curr_liab; prev_wc_val = prev_assets - prev_liab
                 components['wc'] = curr_wc_val - prev_wc_val
                 log_list.append(f"    Change in WC (BS Calc): ({format_currency(curr_assets,'',0)} - {format_currency(curr_liab,'',0)}) - ({format_currency(prev_assets,'',0)} - {format_currency(prev_liab,'',0)}) = {format_currency(components['wc'],'',0)}")
             else: components['wc_src'] = "BS Fail"; log_list.append("    Change in WC (BS Calc): Failed (Missing components)")
        else: components['wc_src'] = "No Prev BS"; log_list.append("    Change in WC (BS Calc): Failed (Missing BS data)")

    # CFO
    cfo_keys = ["Cash Flow From Operating Activities", "Operating Cash Flow", "Total Cash From Operating Activities"]
    components['cfo'] = safe_get(cf_col, cfo_keys, "Cash Flow", default=np.nan)
    if np.isfinite(components['cfo']): log_list.append(f"    CFO: {format_currency(components['cfo'], '', 0)}")
    else: log_list.append(f"    CFO: N/A")

    # Interest Expense
    interest_exp_keys = ['Interest Expense', 'Interest Expense Net']
    components['int_exp'] = abs(safe_get(fin_col, interest_exp_keys, "Income Stmt", default=np.nan))
    if np.isfinite(components['int_exp']): log_list.append(f"    Interest Expense (Abs): {format_currency(components['int_exp'], '', 0)}")
    else: log_list.append(f"    Interest Expense: N/A")

    # Tax Provision
    tax_prov_keys = ['Tax Provision', 'Income Tax Expense']
    components['tax_prov'] = safe_get(fin_col, tax_prov_keys, "Income Stmt", default=np.nan)
    if np.isfinite(components['tax_prov']): log_list.append(f"    Tax Provision: {format_currency(components['tax_prov'], '', 0)}")
    else: log_list.append(f"    Tax Provision: N/A")

    return components

# --- NEW FCFF Calculation Helper: Single Method for Single Year (V21.3) ---
def calculate_fcff_single_method_single_year(method_choice, fin_col, bs_col, cf_col, bs_prev_col, tax_rt, log_list_parent):
    """
    Calculates FCFF for a single year using ONLY the specified method_choice.
    Returns the calculated FCFF value or np.nan if calculation fails.
    Includes detailed logging.
    """
    log_list_local = [] # Local log for component fetching
    c = _get_fcff_components(fin_col, bs_col, cf_col, bs_prev_col, log_list_local)
    log_list_parent.extend(log_list_local) # Append component logs to parent log

    fcff_val = np.nan
    method_name = FCFF_METHODS_DEFINITION.get(method_choice, {}).get("name", "Unknown Method")
    log_list_parent.append(f"--- Calculating FCFF using Selected Method: {method_name} ---")

    # Ensure tax rate is valid if needed
    valid_tax_rate = np.isfinite(tax_rt)
    if not valid_tax_rate and method_choice in [1, 2]: # Methods needing tax rate
        log_list_parent.append(f"    ❌ Cannot calculate {method_name}: Invalid Tax Rate ({tax_rt}).")
        return np.nan

    try:
        # Method 0: EBITDA - TaxProv - WC - Capex
        if method_choice == 0:
            m0_components = [c.get('ebitda'), c.get('tax_prov'), c.get('wc'), c.get('capex')]
            if all(map(np.isfinite, m0_components)):
                fcff_val = c['ebitda'] - c['tax_prov'] - c['wc'] - c['capex']
                log_list_parent.append(f"    FCFF = {format_currency(c['ebitda'],'',0)} - {format_currency(c['tax_prov'],'',0)} - {format_currency(c['wc'],'',0)} - {format_currency(c['capex'],'',0)} = {format_currency(fcff_val,'',0)}")
            else: log_list_parent.append(f"    ❌ Calculation Failed (Missing Components: EBITDA={np.isfinite(m0_components[0])}, TaxProv={np.isfinite(m0_components[1])}, WC={np.isfinite(m0_components[2])}, Capex={np.isfinite(m0_components[3])})")

        # Method 1: NOPAT + Depr - WC - Capex = EBIT*(1-T) + Depr - WC - Capex
        elif method_choice == 1:
            m1_components = [c.get('ebit'), c.get('dep_amort'), c.get('wc'), c.get('capex')]
            if all(map(np.isfinite, m1_components)):
                nopat_m1 = c['ebit'] * (1 - tax_rt)
                fcff_val = nopat_m1 + c['dep_amort'] - c['capex'] - c['wc']
                log_list_parent.append(f"    FCFF = ({format_currency(c['ebit'],'',0)}*(1-{format_percent(tax_rt,1)})) + {format_currency(c['dep_amort'],'',0)} - {format_currency(c['capex'],'',0)} - {format_currency(c['wc'],'',0)} = {format_currency(fcff_val,'',0)}")
            else: log_list_parent.append(f"    ❌ Calculation Failed (Missing Components: EBIT={np.isfinite(m1_components[0])}, Dep={np.isfinite(m1_components[1])}, WC={np.isfinite(m1_components[2])}, Capex={np.isfinite(m1_components[3])})")

        # Method 2: CFO + IntExp*(1-T) - Capex
        elif method_choice == 2:
            m2_components = [c.get('cfo'), c.get('int_exp'), c.get('capex')]
            if all(map(np.isfinite, m2_components)):
                int_tax_shield = c['int_exp'] * (1 - tax_rt)
                fcff_val = c['cfo'] - c['capex'] + int_tax_shield
                log_list_parent.append(f"    FCFF = {format_currency(c['cfo'],'',0)} - {format_currency(c['capex'],'',0)} + ({format_currency(c['int_exp'],'',0)}*(1-{format_percent(tax_rt,1)})) = {format_currency(fcff_val,'',0)}")
            else: log_list_parent.append(f"    ❌ Calculation Failed (Missing Components: CFO={np.isfinite(m2_components[0])}, IntExp={np.isfinite(m2_components[1])}, Capex={np.isfinite(m2_components[2])})")

        # Method 3: CFO - Capex
        elif method_choice == 3:
            m3_components = [c.get('cfo'), c.get('capex')]
            if all(map(np.isfinite, m3_components)):
                fcff_val = c['cfo'] - c['capex']
                log_list_parent.append(f"    FCFF = {format_currency(c['cfo'],'',0)} - {format_currency(c['capex'],'',0)} = {format_currency(fcff_val,'',0)}")
            else: log_list_parent.append(f"    ❌ Calculation Failed (Missing Components: CFO={np.isfinite(m3_components[0])}, Capex={np.isfinite(m3_components[1])})")

        # Method 4: EBITDA - TaxProv (Proxy)
        elif method_choice == 4:
            m4_components = [c.get('ebitda'), c.get('tax_prov')]
            if all(map(np.isfinite, m4_components)):
                fcff_val = c['ebitda'] - c['tax_prov']
                log_list_parent.append(f"    FCFF (Proxy) = {format_currency(c['ebitda'],'',0)} - {format_currency(c['tax_prov'],'',0)} = {format_currency(fcff_val,'',0)}")
                log_list_parent.append(f"    ⚠️ Note: This is a simplified proxy and ignores WC/CapEx changes.")
            else: log_list_parent.append(f"    ❌ Calculation Failed (Missing Components: EBITDA={np.isfinite(m4_components[0])}, TaxProv={np.isfinite(m4_components[1])})")

        else:
            log_list_parent.append(f"    ❌ Error: Invalid method_choice ({method_choice}) provided.")
            return np.nan

        # Check if calculation resulted in non-finite value
        if not np.isfinite(fcff_val):
            log_list_parent.append(f"    ❌ Calculation Result is Non-Finite (NaN/Inf).")
            fcff_val = np.nan # Ensure return is NaN

    except Exception as e:
        log_list_parent.append(f"    ❌ Unexpected Error during FCFF calculation for method {method_choice}: {e}")
        print(traceback.format_exc()) # Print detailed traceback for debugging
        return np.nan

    return fcff_val

# --- REMOVED Helpers (No longer needed): ---
# _calculate_fcff_single_year_best_estimate
# _calculate_all_fcff_bases_v21_4yr
# get_positive_average_fcff_basis_v21_4yr

# --- End Helper Functions ---

# %% [markdown]
# ## 3. User Input & Dynamic Fetching: Ticker, Rf, Rm, Benchmarks, Beta, Spread
# (Section Unchanged)

# %%
input_sources = {}

# --- Get Ticker Symbol ---
user_ticker = input(f"Enter stock ticker (e.g., MSFT, AAPL, RELIANCE.NS) [Default: {DEFAULT_TICKER}]: ")
if not user_ticker: user_ticker = DEFAULT_TICKER
ticker_symbol = user_ticker.strip().upper()
display(Markdown(f"**Analyzing Ticker:** `{ticker_symbol}`"))

# ----> DETECT IF INDIAN STOCK for currency formatting <----
is_indian_stock = ticker_symbol.endswith((".NS", ".BO")) # Include BSE
if is_indian_stock:
    print("Info: Detected Indian stock (.NS or .BO). Using ₹/Lakh/Crore formatting.")

# --- Determine RF Ticker based on Stock Ticker ---
rf_proxy_ticker = "^TNX" # Default US 10Y
if is_indian_stock: rf_proxy_ticker = "^NS10"; print(f"Info: Using Indian Rf proxy: {rf_proxy_ticker}")
elif ticker_symbol.endswith(".L"): rf_proxy_ticker = "^FTLC"; print(f"Info: Using UK Rf proxy: {rf_proxy_ticker}") # Example UK

# --- Get Risk-Free Rate (Rf) ---
print("\n--- CAPM Inputs ---")
dynamic_rf_rate = try_get_dynamic_rf(rf_proxy_ticker)
rf_prompt = f"Enter Risk-Free Rate (Rf) as percentage (e.g., '4.5' for 4.5%). Fetched {rf_proxy_ticker}?"
suggested_rf_val = dynamic_rf_rate * 100 if dynamic_rf_rate is not None else DEFAULT_RF_RATE_SUGGESTION
risk_free_rate_input_pct = get_validated_float_input(rf_prompt, default_value=f"{suggested_rf_val:.2f}", allow_blank=False, min_val=0.0, max_val=20.0, is_percentage=True)
risk_free_rate_input = risk_free_rate_input_pct / 100.0
input_sources['Rf'] = 'Dynamic Fetch (User Confirmed/Adjusted)' if dynamic_rf_rate is not None else f'User Input (Default Suggestion: {DEFAULT_RF_RATE_SUGGESTION:.1f}%)'
display(Markdown(f"**Using Risk-Free Rate (Rf):** {format_percent(risk_free_rate_input)} (`Source: {input_sources['Rf']}`)"))

# --- Get Expected Market Return (Rm) ---
suggested_mrp = DEFAULT_MARKET_RISK_PREMIUM
suggested_rm = risk_free_rate_input + suggested_mrp
rm_prompt = f"Enter Expected Market Return (Rm) as percentage (e.g., '10'). Suggestion based on Rf + {format_percent(suggested_mrp)} MRP is {format_percent(suggested_rm)}"
while True:
    market_return_input_pct = get_validated_float_input(rm_prompt, default_value=f"{suggested_rm * 100:.1f}", allow_blank=False, min_val=-10.0, max_val=35.0, is_percentage=True)
    market_return_input = market_return_input_pct / 100.0
    if market_return_input >= risk_free_rate_input: break
    else: print(f"❌ Error: Rm ({format_percent(market_return_input)}) must be >= Rf ({format_percent(risk_free_rate_input)}). Re-enter.")
market_risk_premium_input = market_return_input - risk_free_rate_input
input_sources['Rm'] = 'User Input'; input_sources['MRP'] = 'Derived'
display(Markdown(f"**Using Expected Market Return (Rm):** {format_percent(market_return_input)} (`Source: {input_sources['Rm']}`)"))
display(Markdown(f"**↳ Implied Market Risk Premium (MRP):** {format_percent(market_risk_premium_input)} (`Source: {input_sources['MRP']}`)"))

# --- Beta (Placeholder) ---
beta_input = None; input_sources['Beta'] = 'Pending'
# --- Credit Spread (Placeholder) ---
user_credit_spread = None; input_sources['Credit Spread'] = 'Pending'

# --- Get Industry Benchmark Ratios (with fallback defaults) ---
print("\n--- Industry Ratio Benchmarks ---")
display(Markdown("Enter benchmarks or leave blank for generic fallbacks. *(Find industry averages online)*"))
user_benchmarks = {}; benchmark_sources = {}
print("Provide benchmarks:")
for ratio_key in REQUIRED_BENCHMARK_RATIOS:
    default_val = DEFAULT_BENCHMARK_RATIOS.get(ratio_key, None)
    prompt = f"  Enter benchmark for {ratio_key}"
    is_percent_type = ratio_key in ['ROE', 'ROA', 'ROCE'] # These should be entered as percentages
    min_b, max_b = (-100, 200) if is_percent_type else (-100, 500)
    bench_val_input = get_validated_float_input(prompt, default_value=f"{default_val:.2f}" if default_val is not None else None, allow_blank=True, min_val=min_b, max_val=max_b, is_percentage=is_percent_type)

    if bench_val_input is not None:
        user_benchmarks[ratio_key] = bench_val_input; benchmark_sources[ratio_key] = 'User Input'
        display_val = f"{bench_val_input:.1f}%" if is_percent_type else f"{bench_val_input:.2f}"
        if ratio_key == 'Asset Turnover': display_val = f"{bench_val_input:.2f}"
        display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: User`)"))
    elif default_val is not None:
         user_benchmarks[ratio_key] = default_val; benchmark_sources[ratio_key] = 'Fallback Default'
         display_val = f"{default_val:.1f}%" if is_percent_type else f"{default_val:.2f}"
         if ratio_key == 'Asset Turnover': display_val = f"{default_val:.2f}"
         display(Markdown(f"  - **{ratio_key} Benchmark:** {display_val} (`Source: Default`)"))
    else:
        user_benchmarks[ratio_key] = np.nan # Store NaN if no input and no default
        benchmark_sources[ratio_key] = 'Not Available'
        display(Markdown(f"  - **{ratio_key} Benchmark:** Not Provided / No Default"))
input_sources['Benchmarks'] = 'User Input or Default'


# %% [markdown]
# ## 4. Data Acquisition (Fetching Extended History)
# (Section Unchanged)

# %%
stock_data = {}
try:
    stock = yf.Ticker(ticker_symbol)
    info = stock.info
    if not info or not info.get('regularMarketPrice'):
        print(f"Warning: Limited info for {ticker_symbol}. Trying history...")
        hist_check = stock.history(period="5d");
        if hist_check.empty: raise ValueError(f"Ticker '{ticker_symbol}' invalid/no data.")
        else: print(f"Warning: Proceeding with limited data."); info = {} if not isinstance(info, dict) else info
    stock_data['info'] = info

    # Fetch statements - use financials/balance_sheet/cashflow
    stock_data['financials'] = stock.financials # Annual usually
    stock_data['balance_sheet'] = stock.balance_sheet
    stock_data['cashflow'] = stock.cashflow

    # Sometimes quarterly data is fetched by default, try forcing annual if empty
    if stock_data['financials'].empty:
        print("Warning: Annual financials empty, trying quarterly...")
        stock_data['financials'] = stock.quarterly_financials
    if stock_data['balance_sheet'].empty:
        print("Warning: Annual balance sheet empty, trying quarterly...")
        stock_data['balance_sheet'] = stock.quarterly_balance_sheet
    if stock_data['cashflow'].empty:
        print("Warning: Annual cashflow empty, trying quarterly...")
        stock_data['cashflow'] = stock.quarterly_cashflow

    # Final Check
    missing_statements = []
    if stock_data['financials'].empty: missing_statements.append("Financials")
    if stock_data['balance_sheet'].empty: missing_statements.append("Balance Sheet")
    if stock_data['cashflow'].empty: missing_statements.append("Cash Flow")
    if missing_statements:
       print(f"⚠️ Warning: Could not retrieve essential statement(s): {', '.join(missing_statements)} for {ticker_symbol}. DCF unreliable/likely to fail.")

    stock_data['history'] = stock.history(period="2y") # For main price chart
    stock_data['history_long'] = stock.history(period="4y", interval="1d") # For stock CAGR calc
    if not stock_data['history_long'].empty:
        print(f"Info: Fetched long history (4y daily). Columns available: {list(stock_data['history_long'].columns)}")
    else:
         print(f"⚠️ Warning: Could not fetch long history (4y daily) for {ticker_symbol}. Stock CAGR calc may fail.")
    if stock_data['history_long'].empty or len(stock_data['history_long']) < 2:
        print(f"⚠️ Warning: Fetched long history has < 2 data points. Stock CAGR calc will fail.")

    company_name = info.get("longName", info.get("shortName", ticker_symbol))
    currency_code = info.get("currency", DEFAULT_CURRENCY_SYMBOL)
    if not isinstance(currency_code, str) or len(currency_code)>5 : currency_code=DEFAULT_CURRENCY_SYMBOL
    currency_symbol = "₹" if is_indian_stock else info.get("symbol", currency_code)

    # Robust price fetching
    current_price = safe_get(info, ['currentPrice', 'regularMarketPrice', 'previousClose', 'lastPrice', 'open', 'dayHigh', 'dayLow'], 'Info', default=np.nan)
    price_source = "yfinance Info"
    if not np.isfinite(current_price) and stock_data['history'] is not None and not stock_data['history'].empty:
        if 'Close' in stock_data['history'].columns and not stock_data['history']['Close'].empty:
            last_close_hist = stock_data['history']['Close'].iloc[-1]
            if pd.notna(last_close_hist) and last_close_hist > 0: current_price = last_close_hist; price_source = "yfinance History (Last Close)"
        elif 'Adj Close' in stock_data['history'].columns and not stock_data['history']['Adj Close'].empty:
             last_adj_close_hist = stock_data['history']['Adj Close'].iloc[-1]
             if pd.notna(last_adj_close_hist) and last_adj_close_hist > 0: current_price = last_adj_close_hist; price_source = "yfinance History (Last Adj Close)"

    if not np.isfinite(current_price): print(f"⚠️ Warning: Could not determine current price."); current_price = np.nan

    shares_outstanding = safe_get(info, ['sharesOutstanding'], 'Info', default=0)
    if not np.isfinite(shares_outstanding) or shares_outstanding <= 0:
        print("Warning: Shares outstanding from info is zero/invalid. Trying Balance Sheet...")
        if stock_data['balance_sheet'] is not None and not stock_data['balance_sheet'].empty:
            shares_keys_bs = ['Share Issued', 'Ordinary Shares Number', 'Common Stock Equity']
            shares_bs = safe_get(stock_data['balance_sheet'].iloc[:,0], shares_keys_bs, 'Balance Sheet', default=0)
            if shares_bs is not None and np.isfinite(shares_bs) and shares_bs > 0:
                 shares_outstanding = shares_bs
                 print(f"Info: Using shares outstanding from Balance Sheet: {shares_outstanding:,.0f}")
            else:
                 print("Warning: Could not find valid shares outstanding on Balance Sheet either."); shares_outstanding = 0
        else: print("Warning: Balance sheet empty or invalid, cannot check for shares outstanding."); shares_outstanding = 0
        shares_outstanding = 0 if not np.isfinite(shares_outstanding) else shares_outstanding

    beta_input = safe_get(info, ['beta', 'beta3Year'], 'Info', default=None)
    if beta_input is not None and np.isfinite(beta_input):
        input_sources['Beta'] = 'Dynamic Fetch'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))
    else:
         print(f"⚠️ WARNING: Beta not found dynamically or invalid.")
         beta_input_user = get_validated_float_input(f"Enter Beta for {ticker_symbol} (or leave blank to use 1.0):", default_value="1.0", allow_blank=True, min_val=-1.0, max_val=4.0)
         beta_input = float(beta_input_user) if beta_input_user is not None else 1.0
         info['beta'] = beta_input
         input_sources['Beta'] = 'User Input' if beta_input_user is not None else 'Default (1.0)'
         display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))


    market_cap = safe_get(info, ['marketCap'], 'Info', default=0)
    mcap_source = "yfinance Info"
    if (not np.isfinite(market_cap) or market_cap <= 0) and shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Calculated (Price * Shares)"; print(f"Info: Market Cap (re)calculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}")
    elif not np.isfinite(market_cap) or market_cap <=0: mcap_source = "Unavailable"; print(f"⚠️ Warning: Market Cap unavailable."); market_cap = np.nan

    sector = info.get("sector", "N/A"); industry = info.get("industry", "N/A")
    summary = info.get("longBusinessSummary", "No summary available.")

    display(Markdown(f"**Company:** {company_name}")); display(Markdown(f"**Sector:** {sector} | **Industry:** {industry}"))
    display(Markdown(f"**Currency:** {currency_code} ({currency_symbol})"));
    display(Markdown(f"**Current Market Price:** {format_currency(current_price, currency_symbol, is_indian=is_indian_stock)} (`Source: {price_source}`)"))
    display(Markdown(f"**Market Cap:** {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)} (`Source: {mcap_source}`)"))
    display(Markdown(f"**Shares Outstanding:** {shares_outstanding:,.0f}"))


except Exception as e:
    print(f"❌ CRITICAL ERROR fetching data for {ticker_symbol}: {e}")
    print(traceback.format_exc())
    sys.exit("Cannot proceed due to data fetching error.")

# Essential Data Checks After Fetching
if not isinstance(shares_outstanding, (int, float)) or shares_outstanding <= 0:
    shares_outstanding_input = get_validated_float_input(f"CRITICAL: Shares Outstanding invalid ({shares_outstanding}). Enter value:", allow_blank=False, min_val=1)
    shares_outstanding = shares_outstanding_input
    if 'info' in stock_data and isinstance(stock_data['info'], dict): stock_data['info']['sharesOutstanding'] = shares_outstanding
    print(f"Info: Using user-provided shares outstanding: {shares_outstanding:,.0f}")

if beta_input is None or not np.isfinite(beta_input):
    beta_input = 1.0 # Default to 1.0 if it somehow became invalid
    print("CRITICAL WARNING: Beta became invalid after fetch/input. Defaulting to 1.0. Review WACC.")
    input_sources['Beta'] = 'Default (1.0) - Invalid Fallback'

if not np.isfinite(market_cap):
    print(f"⚠️ WARNING: Market Cap unavailable/invalid. WACC weights will use Book Equity fallback or placeholder.")
    if shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
       market_cap = shares_outstanding * current_price; info['marketCap'] = market_cap
       mcap_source = "Recalculated"; print(f"Info: Market Cap recalculated = {format_currency(market_cap, currency_symbol, is_indian=is_indian_stock)}")
    else: market_cap = np.nan # Ensure it's NaN

if not np.isfinite(current_price):
    print(f"⚠️ WARNING: Current Price unavailable/invalid. Valuation comparison and some ratios will be N/A.")


# %% [markdown]
# ## 5. WACC Calculation
# (Section Unchanged)

# %%
# WACC Function Definition
def calculate_wacc(stock_info_dict, financials_df, balance_sheet_df,
                   user_rf, user_mrp, user_beta, input_sources_dict,
                   currency_sym, is_indian):
    global user_credit_spread
    display(Markdown("### Calculating WACC..."))
    components = {}
    calculation_log = []
    calculation_log.append(f"Input Sources: Rf='{input_sources_dict.get('Rf')}', MRP='{input_sources_dict.get('MRP')}', Beta='{input_sources_dict.get('Beta')}'")

    # Cost of Equity
    beta_val = user_beta
    if beta_val <= 0: calculation_log.append(f"⚠️ Warning: Non-positive Beta ({beta_val:.2f}) used in CAPM.")
    components['beta'] = beta_val; components['risk_free_rate'] = user_rf; components['market_risk_premium'] = user_mrp
    cost_of_equity = user_rf + beta_val * user_mrp
    if not (0 < cost_of_equity < 0.50): calculation_log.append(f"⚠️ Warning: Calculated Cost of Equity ({format_percent(cost_of_equity)}) seems unusual.")
    components['cost_of_equity'] = cost_of_equity
    calculation_log.append(f"Inputs -> Rf: {format_percent(user_rf)}, Beta: {beta_val:.2f}, MRP: {format_percent(user_mrp)}")
    calculation_log.append(f"↳ Cost of Equity (Ke) [CAPM]: {format_percent(cost_of_equity)}")

    # Cost of Debt & Tax Rate
    cost_of_debt_pre_tax = np.nan; tax_rate = np.nan; after_tax_cost_of_debt = np.nan
    total_debt = 0.0; debt_calculation_method = "N/A"; tax_rate_source = "N/A"
    statements_available = (financials_df is not None and not financials_df.empty and
                            balance_sheet_df is not None and not balance_sheet_df.empty)

    if statements_available:
        latest_bs = balance_sheet_df.iloc[:, 0]; latest_fin = financials_df.iloc[:, 0]
        # Total Debt
        debt_keys_long = ['Total Debt Non Current', 'Long Term Debt', 'Long Term Debt And Capital Lease Obligation']
        debt_keys_short = ['Total Debt Current', 'Current Debt', 'Short Term Debt', 'Current Debt And Capital Lease Obligation', 'Short Long Term Debt']
        long_term_debt = safe_get(latest_bs, debt_keys_long , 'Balance Sheet (Long Debt)', default=0)
        short_term_debt = safe_get(latest_bs, debt_keys_short , 'Balance Sheet (Short Debt)', default=0)
        long_term_debt = float(long_term_debt) if np.isfinite(long_term_debt) else 0.0
        short_term_debt = float(short_term_debt) if np.isfinite(short_term_debt) else 0.0
        total_debt = long_term_debt + short_term_debt
        components['debt_value_used'] = total_debt
        calculation_log.append(f"Latest BS Debt -> Long Term: {format_currency(long_term_debt, '', 0, is_indian)}, Short Term: {format_currency(short_term_debt, '', 0, is_indian)} -> Total Debt: {format_currency(total_debt, '', 0, is_indian)}")

        # Pre-Tax Cost of Debt (Kd)
        interest_expense_keys = ['Interest Expense', 'Interest Expense Net']
        interest_expense = abs(safe_get(latest_fin, interest_expense_keys, 'Income Statement (Interest)', default=0))
        interest_expense = float(interest_expense) if np.isfinite(interest_expense) else 0.0
        calculation_log.append(f"Latest Income Stmt -> Interest Expense: {format_currency(interest_expense, '', 0, is_indian)}")

        implied_rate_calculated = False
        if total_debt > 1000 and interest_expense > 0:
             implied_rate = interest_expense / total_debt; implied_rate_calculated = True
             calculation_log.append(f"Implied Kd = Interest / Debt = {format_percent(implied_rate)}")
             if MIN_COST_OF_DEBT <= implied_rate <= MAX_COST_OF_DEBT:
                  cost_of_debt_pre_tax = implied_rate; debt_calculation_method = "Implied (Interest/Debt)";
                  calculation_log.append(f"↳ Using Implied Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)} (Within Bounds)")
             else: calculation_log.append(f"⚠️ Implied Kd out of bounds ({format_percent(MIN_COST_OF_DEBT)}-{format_percent(MAX_COST_OF_DEBT)}). Falling back."); implied_rate_calculated=False
        elif total_debt > 1000: calculation_log.append(f"ℹ️ Cannot use implied Kd (Debt>0 but Interest=0). Falling back.")
        else: calculation_log.append(f"ℹ️ Cannot use implied Kd (Negligible/Zero Debt).")

        if not implied_rate_calculated:
             if total_debt > 1000:
                 debt_calculation_method = "Default (Rf+Spread)"
                 if user_credit_spread is None:
                     spread_prompt = f"Implied Kd failed/NA for Debt={format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd"
                     user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=15.0, is_percentage=True)
                     user_credit_spread = user_credit_spread_pct / 100.0
                     input_sources_dict['Credit Spread'] = 'User Input (for Default Kd)'
                 else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
                 cost_of_debt_pre_tax = user_rf + user_credit_spread
                 original_default_kd = cost_of_debt_pre_tax
                 cost_of_debt_pre_tax = min(max(cost_of_debt_pre_tax, MIN_COST_OF_DEBT), MAX_COST_OF_DEBT)
                 calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(original_default_kd)}")
                 if abs(cost_of_debt_pre_tax - original_default_kd) > 1e-6 : calculation_log.append(f"↳ Bounded Default Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)}")
                 else: calculation_log.append(f"↳ Using Default Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)}")
                 calculation_log.append(f"   (Spread Source: {input_sources_dict['Credit Spread']})")
             else:
                 debt_calculation_method = "Zero Debt"; cost_of_debt_pre_tax = 0.0; input_sources_dict['Credit Spread'] = 'N/A (Zero Debt)';
                 calculation_log.append(f"ℹ️ Negligible/Zero Debt. Setting Kd = 0%.")

        # Tax Rate Calculation
        tax_prov_keys = ['Tax Provision', 'Income Tax Expense']
        pretax_inc_keys = ['Pretax Income', 'EBT', 'Earnings Before Tax']
        income_tax_expense = safe_get(latest_fin, tax_prov_keys, 'Income Statement (Tax Exp)', default=0)
        pretax_income = safe_get(latest_fin, pretax_inc_keys, 'Income Statement (Pretax Inc)', default=0)
        income_tax_expense = float(income_tax_expense) if np.isfinite(income_tax_expense) else 0.0
        pretax_income = float(pretax_income) if np.isfinite(pretax_income) else 0.0
        calculation_log.append(f"Latest Income Stmt -> Tax Expense: {format_currency(income_tax_expense, '', 0, is_indian)}, Pretax Income: {format_currency(pretax_income, '', 0, is_indian)}")

        if pretax_income <= 1 and income_tax_expense > 0:
             ebit_val = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=0);
             ebit_val = float(ebit_val) if np.isfinite(ebit_val) else 0.0
             pretax_income_from_ebit = ebit_val - interest_expense
             if pretax_income_from_ebit > 1:
                 pretax_income = pretax_income_from_ebit
                 calculation_log.append(f"ℹ️ Pretax Income was near zero/negative. Used EBIT-Interest as proxy: {format_currency(pretax_income, symbol=currency_sym, is_indian=is_indian)}")

        if pretax_income > 1 and income_tax_expense >= 0:
            calculated_tax_rate = income_tax_expense / pretax_income;
            tax_rate = min(max(calculated_tax_rate, MIN_TAX_RATE), MAX_TAX_RATE); tax_rate_source = "Derived & Bounded"
            if abs(tax_rate - calculated_tax_rate) > 0.001:
                 tax_rate_source = f"Derived ({format_percent(calculated_tax_rate)}), Bounded"
                 calculation_log.append(f"Calculated Tax Rate = Tax Exp / Pretax Inc = {format_percent(calculated_tax_rate)}")
            calculation_log.append(f"↳ Effective Tax Rate: {format_percent(tax_rate)} ({tax_rate_source})")
        else:
             tax_rate = DEFAULT_TAX_RATE; tax_rate_source = "Default (Derived Failed)";
             calculation_log.append(f"⚠️ Cannot derive tax rate (Pretax Inc <= 0 or Tax Exp < 0). Using default: {format_percent(tax_rate)}")

    else: # Statements unavailable
         calculation_log.append("⚠️ Financial statements empty/invalid. Using defaults for Kd and Tax Rate."); debt_calculation_method = "Default (No Fin.)"; tax_rate_source = "Default (No Fin.)"
         if user_credit_spread is None:
              spread_prompt = f"No financials. Enter Credit Spread over Rf ({format_percent(user_rf)}) for Kd"
              user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=15.0, is_percentage=True)
              user_credit_spread = user_credit_spread_pct / 100.0; input_sources_dict['Credit Spread'] = 'User Input (No Fin.)'
         else: input_sources_dict['Credit Spread'] = 'User Input (Reused)'
         cost_of_debt_pre_tax = user_rf + user_credit_spread;
         original_default_kd = cost_of_debt_pre_tax
         cost_of_debt_pre_tax = min(max(cost_of_debt_pre_tax, MIN_COST_OF_DEBT), MAX_COST_OF_DEBT)
         tax_rate = DEFAULT_TAX_RATE; total_debt = 0; components['debt_value_used'] = 0
         calculation_log.append(f"Using Default Kd: Rf ({format_percent(user_rf)}) + Spread ({format_percent(user_credit_spread)}) = {format_percent(original_default_kd)}")
         if abs(cost_of_debt_pre_tax - original_default_kd) > 1e-6 : calculation_log.append(f"↳ Bounded Default Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)}")
         else: calculation_log.append(f"↳ Using Default Cost of Debt (pre-tax): {format_percent(cost_of_debt_pre_tax)}")
         calculation_log.append(f"Using Default Tax Rate: {format_percent(tax_rate)}")

    # After-tax Cost of Debt
    components['cost_of_debt'] = cost_of_debt_pre_tax if np.isfinite(cost_of_debt_pre_tax) else 0.0
    components['debt_calc_method'] = debt_calculation_method
    components['tax_rate'] = tax_rate if np.isfinite(tax_rate) else DEFAULT_TAX_RATE
    components['tax_rate_source'] = tax_rate_source

    if np.isfinite(components['cost_of_debt']) and np.isfinite(components['tax_rate']):
        after_tax_cost_of_debt = components['cost_of_debt'] * (1 - components['tax_rate'])
    else:
        calculation_log.append(f"⚠️ Invalid Kd ({components['cost_of_debt']}) or Tax Rate ({components['tax_rate']}). Cannot calculate after-tax Kd.")
        after_tax_cost_of_debt = np.nan

    components['after_tax_cost_of_debt'] = after_tax_cost_of_debt
    if np.isfinite(after_tax_cost_of_debt) and components['cost_of_debt'] > 0 : calculation_log.append(f"↳ Cost of Debt (After-tax): {format_percent(after_tax_cost_of_debt)}")

    # Equity and Debt Weights
    market_cap_val = safe_get(stock_info_dict, ['marketCap'], 'Info', default=np.nan); equity_source = "Market Cap" if np.isfinite(market_cap_val) and market_cap_val > 0 else "N/A"
    equity_value_for_wacc = np.nan

    if np.isfinite(market_cap_val) and market_cap_val > 0: equity_value_for_wacc = market_cap_val; equity_source = "Market Cap"
    else:
       calculation_log.append(f"⚠️ Market Cap unavailable/zero ({format_currency(market_cap_val, symbol=currency_sym, is_indian=is_indian)}). Falling back to Book Equity."); equity_source = "Book Equity Fallback"
       if balance_sheet_df is not None and not balance_sheet_df.empty:
           equity_keys = ['Total Stockholder Equity', 'Stockholders Equity', 'Total Equity Gross Minority Interest', 'Common Stock Equity']
           book_equity_val = safe_get(balance_sheet_df.iloc[:, 0], equity_keys, 'Balance Sheet (Book Equity)', default=np.nan)
           if np.isfinite(book_equity_val) and book_equity_val > 1.0:
              equity_value_for_wacc = book_equity_val;
              calculation_log.append(f"Using Book Equity ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}) as proxy.")
           else: equity_source = "Placeholder (Book Invalid)"; calculation_log.append(f"⚠️ Book Equity also invalid/unavailable ({format_currency(book_equity_val, symbol=currency_sym, is_indian=is_indian)}).")
       else: equity_source = "Placeholder (No BS)"
       if not np.isfinite(equity_value_for_wacc):
            equity_value_for_wacc = 1.0; equity_source = "Placeholder (Market/Book Failed)";
            calculation_log.append("⚠️ Market Cap & Book Equity unavailable/invalid. Using placeholder Equity Value (1.0). WACC Weights inaccurate.")

    components['equity_value_used'] = equity_value_for_wacc; components['equity_source'] = equity_source
    equity_weight = np.nan; debt_weight = np.nan; weights_source = "N/A";
    total_capital = equity_value_for_wacc + total_debt
    if np.isfinite(total_capital) and total_capital > 1.0:
        equity_weight = equity_value_for_wacc / total_capital; debt_weight = total_debt / total_capital; weights_source = "Calculated"
        total_weight = equity_weight + debt_weight
        if abs(total_weight - 1.0) > 1e-6:
             calculation_log.append(f"ℹ️ Normalizing E/D weights (Sum was {total_weight:.4f})")
             equity_weight /= total_weight; debt_weight /= total_weight
    else: calculation_log.append(f"❌ ERROR: Total Capital ({format_currency(total_capital, symbol=currency_sym, is_indian=is_indian)}) invalid/zero. Using default weights (70/30)."); equity_weight = 0.7; debt_weight = 0.3; weights_source = "Default (Invalid Cap)"

    components['equity_weight'] = equity_weight; components['debt_weight'] = debt_weight; components['weights_source'] = weights_source
    calculation_log.append(f"Equity Value Used ({components['equity_source']}): {format_currency(components['equity_value_used'], symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"Debt Value Used ({components['debt_calc_method']}): {format_currency(total_debt, symbol=currency_sym, is_indian=is_indian)}")
    calculation_log.append(f"↳ Weights E/V: {format_percent(equity_weight)}, D/V: {format_percent(debt_weight)} (Source: {weights_source})")

    # Final WACC Calculation
    wacc = np.nan
    if all(map(np.isfinite, [equity_weight, cost_of_equity, debt_weight, after_tax_cost_of_debt])):
        wacc = (equity_weight * cost_of_equity) + (debt_weight * after_tax_cost_of_debt); original_wacc = wacc;
        wacc = min(max(wacc, MIN_WACC), MAX_WACC)
        bound_note = ""
        if abs(wacc - original_wacc) > 0.0001 : bound_note = f"(Bounded from {format_percent(original_wacc)})"; calculation_log.append(f"⚠️ WACC bounded to {format_percent(MIN_WACC)}-{format_percent(MAX_WACC)}.")
        calculation_log.append(f"▶️ Calculated WACC: {format_percent(wacc)} {bound_note}")
    else:
        calculation_log.append(f"❌ ERROR: Cannot calculate final WACC due to invalid components.")
        wacc = np.nan

    components['final_wacc'] = wacc
    display(Markdown("```\n" + "\n".join(calculation_log) + "\n```"))
    # Plot WACC Donut Chart
    if np.isfinite(wacc) and wacc > 0 and np.isfinite(equity_weight) and np.isfinite(debt_weight):
        equity_contrib = equity_weight * cost_of_equity; debt_contrib = debt_weight * after_tax_cost_of_debt
        if np.isfinite(equity_contrib) and np.isfinite(debt_contrib) and abs(equity_contrib + debt_contrib - wacc) < 0.01:
           labels = ['Cost of Equity Contribution', 'Cost of Debt Contribution']; values = [equity_contrib, debt_contrib]
           fig_wacc_donut = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4, marker_colors=px.colors.qualitative.Pastel, hovertemplate='<b>%{label}</b><br>Contribution: %{value:.2%}<extra></extra>', textinfo='percent', textfont_size=12, sort=False)])
           fig_wacc_donut.update_layout(title_text=f'WACC Composition (Total: {format_percent(wacc)})', title_x=0.5, showlegend=True, legend_title_text='Components', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), margin=dict(l=20, r=20, t=60, b=50), height=350)
           fig_wacc_donut.show()
        else: display(Markdown(f"*(WACC component contributions invalid/don't sum. Cannot generate composition chart.)*"))
    elif not np.isfinite(wacc): display(Markdown(f"*(WACC calculation failed. Cannot generate composition chart)*"))
    else: display(Markdown(f"*(WACC zero/negative or weights invalid. Cannot generate composition chart)*"))

    return wacc, components

# Execute WACC Calculation
discount_rate, wacc_components = calculate_wacc(
    stock_data['info'], stock_data['financials'], stock_data['balance_sheet'],
    risk_free_rate_input, market_risk_premium_input, beta_input, input_sources,
    currency_symbol, is_indian_stock
)

# %% [markdown]
# ## 6. User FCFF Method Selection

# %%
# --- Get User Choice for FCFF Method ---
selected_fcff_method = None
print("\n--- FCFF Calculation Method Selection ---")
print("Please choose the FCFF calculation method to use for historical data, CAGR, and projection basis:")

# Display options
for method_key, method_info in FCFF_METHODS_DEFINITION.items():
    print(f"  {method_key}. {method_info['name']}")
    print(f"     Formula: {method_info['formula']}")

# Get validated input
while selected_fcff_method is None:
    try:
        choice = input(f"Enter the number of the desired method [{', '.join(map(str, FCFF_METHODS_DEFINITION.keys()))}]: ")
        method_idx = int(choice)
        if method_idx in FCFF_METHODS_DEFINITION:
            selected_fcff_method = method_idx
            method_name_chosen = FCFF_METHODS_DEFINITION[selected_fcff_method]['name']
            display(Markdown(f"**Selected FCFF Method:** `{method_idx} - {method_name_chosen}`"))
            input_sources['FCFF Method'] = f"User Choice ({method_name_chosen})"
        else:
            print(f"❌ Invalid choice. Please enter a number from {list(FCFF_METHODS_DEFINITION.keys())}.")
    except ValueError:
        print("❌ Invalid input. Please enter a number.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        sys.exit("Halting due to input error.")


# %% [markdown]
# ## 7. FCFF Calculation & Projection (V21.3 - Single User Method / 3yr)

# %%
# FCFF Function Definition (V21.3 - Uses Single User Method)
def calculate_and_project_fcff_user_selected_method_v21_3(
                                   user_selected_method_idx, # User's choice (0-4)
                                   financials_df, balance_sheet_df, cashflow_df, history_long_df,
                                   wacc_comps, disc_rate, forecast_yrs, # forecast_yrs is now 3
                                   final_user_rf, default_init_growth,
                                   currency_sym, is_indian):

    display(Markdown(f"### Calculating & Projecting FCFF (Using Method: {FCFF_METHODS_DEFINITION[user_selected_method_idx]['name']})"))
    projection_log = [] # Master log for this section
    projection_results = {
        'historical_fcffs_selected_method': {}, # Store T0-T3 FCFF using selected method
        'historical_revenues': {}, # Store recent historical revenues
        'current_fcff_basis': np.nan, # This will be the 4yr average of selected method's FCFF
        'projected_fcffs': [], 'growth_rates': [],
        'pv_fcffs': [], 'cumulative_pv_fcff': np.nan, 'terminal_growth_rate': np.nan,
        'terminal_value': np.nan, 'pv_terminal_value': np.nan, 'error': None,
        'initial_growth_rate': np.nan, 'growth_source': 'N/A',
        'fcff_basis_method': f"{FCFF_METHODS_DEFINITION[user_selected_method_idx]['name']} (User Selected)"
    }

    # --- Pre-checks ---
    statements_valid = True
    if financials_df is None or financials_df.empty:
        projection_results['error'] = "Missing Financials Statement."; statements_valid = False
    if balance_sheet_df is None or balance_sheet_df.empty:
        projection_results['error'] = (projection_results['error'] + " Missing Balance Sheet.") if projection_results['error'] else "Missing Balance Sheet."
        statements_valid = False
    if cashflow_df is None or cashflow_df.empty:
        projection_results['error'] = (projection_results['error'] + " Missing Cash Flow.") if projection_results['error'] else "Missing Cash Flow."
        statements_valid = False

    if not statements_valid:
         projection_log.append(f"❌ ERROR: {projection_results['error']} Cannot proceed.")
         display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
         return projection_results

    if not np.isfinite(disc_rate):
         projection_results['error'] = "Invalid WACC provided.";
         projection_log.append(f"❌ ERROR: {projection_results['error']} Cannot proceed.")
         display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
         return projection_results

    # --- Get Effective Tax Rate ---
    effective_tax_rate = DEFAULT_TAX_RATE; tax_rate_log_src = "Default"
    if isinstance(wacc_comps, dict) and 'tax_rate' in wacc_comps and np.isfinite(wacc_comps['tax_rate']):
        effective_tax_rate = wacc_comps['tax_rate']
        tax_rate_log_src = wacc_comps.get('tax_rate_source', '?')
    projection_log.append(f"Using Tax Rate from WACC: {format_percent(effective_tax_rate)} ({tax_rate_log_src})")

    # ====================================================================
    # PHASE 1: Calculate Historical Data (Using ONLY Selected FCFF Method)
    # ====================================================================
    projection_log.append(f"\n--- Phase 1: Calculating Historical FCFF & Revenue (using Selected Method {user_selected_method_idx}) ---")
    projection_log.append(f"    (Method: {FCFF_METHODS_DEFINITION[user_selected_method_idx]['name']})")

    num_hist_years_needed = 4 # Need T-0, T-1, T-2, T-3 for 4yr avg and 3yr CAGR
    num_statement_years_avail = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    historical_fcffs_list_selected = [] # FCFF using ONLY selected method
    historical_revenues_list = [] # Newest first (index 0 = T-0)

    years_to_calc_hist = min(num_statement_years_avail, num_hist_years_needed)
    if years_to_calc_hist < num_hist_years_needed:
         projection_log.append(f"  ⚠️ Warning: Only {years_to_calc_hist} years of statement data available (needed {num_hist_years_needed} for full 4yr avg / 3yr CAGR). Results may be unreliable/fail.")

    for i in range(years_to_calc_hist):
        projection_log.append(f"\n--- Processing Historical Year T-{i} ---")
        # Check if data exists for this specific year 'i'
        if i >= financials_df.shape[1] or i >= balance_sheet_df.shape[1] or i >= cashflow_df.shape[1]:
            projection_log.append(f"  --> Historical FCFF T-{i} (Selected Method): N/A (Data Missing)")
            projection_log.append(f"  --> Historical Revenue T-{i}: N/A (Data Missing)")
            historical_fcffs_list_selected.append(np.nan)
            historical_revenues_list.append(np.nan)
            continue

        fin_col = financials_df.iloc[:, i]; bs_col = balance_sheet_df.iloc[:, i]; cf_col = cashflow_df.iloc[:, i]
        # Need previous year's BS for WC calculation in some methods
        bs_prev_col = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None
        if bs_prev_col is None and user_selected_method_idx in [0, 1]: # Methods potentially needing BS WC calc
             projection_log.append(f"    ⚠️ Warning: Previous Balance Sheet data unavailable for T-{i}. WC calculation might fail or use CF fallback.")

        # Calculate FCFF using ONLY the selected method
        fcff_val = calculate_fcff_single_method_single_year(
            user_selected_method_idx, fin_col, bs_col, cf_col, bs_prev_col, effective_tax_rate, projection_log
        )
        historical_fcffs_list_selected.append(fcff_val)
        projection_log.append(f"  --> Historical FCFF T-{i} (Selected Method {user_selected_method_idx}): {format_currency(fcff_val, '', 0)}")

        # Get Revenue for CAGR calculation
        rev_keys = ["Total Revenue", "Revenue", "Net Revenue"]
        revenue_val = safe_get(fin_col, rev_keys, "Income Stmt", default=np.nan)
        historical_revenues_list.append(revenue_val)
        projection_log.append(f"  --> Historical Revenue T-{i}: {format_currency(revenue_val, '', 0)}")

    # Pad with NaNs if data is shorter than needed
    while len(historical_fcffs_list_selected) < num_hist_years_needed: historical_fcffs_list_selected.append(np.nan)
    while len(historical_revenues_list) < num_hist_years_needed: historical_revenues_list.append(np.nan)

    projection_results['historical_fcffs_selected_method'] = {f"T-{i}": v for i, v in enumerate(historical_fcffs_list_selected)}
    projection_results['historical_revenues'] = {f"T-{i}": v for i, v in enumerate(historical_revenues_list)}

    # =====================================================================
    # PHASE 2: Determine & Check Projection Basis (4yr Avg of Selected)
    # =====================================================================
    projection_log.append(f"\n--- Phase 2: Determining Projection Basis (4yr Avg using Selected Method {user_selected_method_idx}) ---")
    fcff_start_basis = np.nan
    valid_fcff_count = sum(1 for f in historical_fcffs_list_selected if np.isfinite(f))

    if valid_fcff_count > 0:
        # Calculate average of the historical FCFFs from the selected method
        avg_fcff_selected = np.nanmean(historical_fcffs_list_selected) # nanmean handles potential NaNs
        projection_log.append(f"  Historical FCFFs (Selected Method): { {k: format_currency(v,'',0) for k,v in projection_results['historical_fcffs_selected_method'].items()} }")
        projection_log.append(f"  Average of {valid_fcff_count} valid year(s): {format_currency(avg_fcff_selected, '', 0)}")

        # Check if the average is positive
        if np.isfinite(avg_fcff_selected) and avg_fcff_selected > 1e-9:
            fcff_start_basis = avg_fcff_selected
            projection_log.append(f"  ✅ Average is positive. Using this as the projection starting basis.")
            projection_results['current_fcff_basis'] = fcff_start_basis
            # fcff_basis_method already set during user selection
        else:
            projection_results['error'] = f"4yr Average FCFF for selected method ({FCFF_METHODS_DEFINITION[user_selected_method_idx]['name']}) is non-positive or invalid ({format_currency(avg_fcff_selected,'',0)}). Cannot proceed."
            projection_log.append(f"  ❌ {projection_results['error']}")
    else:
        projection_results['error'] = f"No valid historical FCFF values calculated using selected method ({FCFF_METHODS_DEFINITION[user_selected_method_idx]['name']}). Cannot determine average basis."
        projection_log.append(f"  ❌ {projection_results['error']}")

    # --- Halt if basis is not positive ---
    if not np.isfinite(fcff_start_basis):
        projection_log.append(f"\n❌ HALTING: No positive FCFF projection basis determined.")
        display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
        return projection_results # Stop before growth calculation

    # ================================================================
    # PHASE 3: Estimate Growth Rates (Using Data from Ph 1)
    # ================================================================
    projection_log.append("\n--- Phase 3: Estimating Initial Growth Rate (CAGR Geometric Mean) ---")
    initial_growth_rate = np.nan # Start with NaN
    growth_source = "N/A"
    calculated_cagrs = {} # Store individual CAGRs {Name: value}
    valid_cagr_list = [] # Store only the valid CAGRs for GM calculation
    num_years_cagr = 3 # Still use 3yr CAGR (T0 vs T3)

    # 1. Calculate 3-Year Revenue CAGR (Conditional)
    rev_cagr = np.nan
    projection_log.append(f"\nCalculating Revenue {num_years_cagr}yr CAGR (T-0 vs T-{num_years_cagr})...")
    if len(historical_revenues_list) >= num_hist_years_needed:
        rev_end = historical_revenues_list[0]; rev_start = historical_revenues_list[num_years_cagr] # T-3
        projection_log.append(f"  Data Points -> Start (T-{num_years_cagr}): {format_currency(rev_start, '', 0, is_indian)}, End (T-0): {format_currency(rev_end, '', 0, is_indian)}")
        if pd.notna(rev_end) and math.isfinite(rev_end) and pd.notna(rev_start) and math.isfinite(rev_start) and rev_start > 1e-9:
            rev_cagr = calculate_cagr_modified(rev_start, rev_end, num_years_cagr)
            if pd.notna(rev_cagr) and math.isfinite(rev_cagr):
                 calculated_cagrs['Revenue'] = rev_cagr; valid_cagr_list.append(rev_cagr)
                 projection_log.append(f"  ✅ Revenue CAGR: {format_percent(rev_cagr)}")
            else: projection_log.append(f"  ❌ Revenue CAGR Calculation Failed (Result Invalid: {rev_cagr}).")
        elif not (pd.notna(rev_start) and math.isfinite(rev_start) and rev_start > 1e-9):
             projection_log.append(f"  ❌ Revenue CAGR Skipped (Invalid/Non-positive Start Data).")
        else: projection_log.append(f"  ❌ Revenue CAGR Skipped (Invalid End Data).")
    else: projection_log.append(f"  ❌ Revenue CAGR Skipped (Insufficient data points).")

    # 2. Calculate 3-Year FCFF CAGR (Using ONLY selected method's history)
    fcff_cagr = np.nan
    projection_log.append(f"\nCalculating FCFF {num_years_cagr}yr CAGR (T-0 vs T-{num_years_cagr}) using Selected Method {user_selected_method_idx}...")
    if len(historical_fcffs_list_selected) >= num_hist_years_needed:
        fcff_end = historical_fcffs_list_selected[0]; fcff_start = historical_fcffs_list_selected[num_years_cagr] # T-3
        projection_log.append(f"  Data Points -> Start (T-{num_years_cagr}): {format_currency(fcff_start, '', 0, is_indian)}, End (T-0): {format_currency(fcff_end, '', 0, is_indian)}")
        if pd.notna(fcff_end) and math.isfinite(fcff_end) and pd.notna(fcff_start) and math.isfinite(fcff_start) and fcff_start > 1e-9:
            fcff_cagr = calculate_cagr_modified(fcff_start, fcff_end, num_years_cagr)
            if pd.notna(fcff_cagr) and math.isfinite(fcff_cagr):
                 calculated_cagrs['FCFF (Selected Method)'] = fcff_cagr; valid_cagr_list.append(fcff_cagr)
                 projection_log.append(f"  ✅ FCFF CAGR:    {format_percent(fcff_cagr)}")
            else: projection_log.append(f"  ❌ FCFF CAGR Calculation Failed (Result Invalid: {fcff_cagr}).")
        elif not (pd.notna(fcff_start) and math.isfinite(fcff_start) and fcff_start > 1e-9):
            projection_log.append(f"  ❌ FCFF CAGR Skipped (Invalid/Non-positive Start Data).")
        else: projection_log.append(f"  ❌ FCFF CAGR Skipped (Invalid End Data).")
    else: projection_log.append(f"  ❌ FCFF CAGR Skipped (Insufficient data points).")

    # 3. Calculate 3-Year Stock Price CAGR (Conditional)
    stock_cagr = np.nan; price_col = None
    projection_log.append(f"\nCalculating Stock Price {num_years_cagr}yr CAGR...")
    # (Stock CAGR logic remains the same as previous version)
    if history_long_df is not None and not history_long_df.empty and len(history_long_df) >= 2:
         if 'Adj Close' in history_long_df.columns and history_long_df['Adj Close'].notna().any(): price_col = 'Adj Close'
         elif 'Close' in history_long_df.columns and history_long_df['Close'].notna().any(): price_col = 'Close'; projection_log.append("  ⚠️ Using 'Close' price for Stock CAGR ('Adj Close' not found or all NaN).")
         else: projection_log.append(f"  ❌ Stock CAGR Skipped (No valid 'Adj Close' or 'Close' columns)")

         if price_col and isinstance(history_long_df.index, pd.DatetimeIndex):
             projection_log.append(f"  Stock History Info: Using '{price_col}'. Range: {history_long_df.index.min().date()} to {history_long_df.index.max().date()}")
             try:
                 history_long_df = history_long_df.sort_index()
                 end_series = history_long_df[price_col].dropna();
                 if end_series.empty: raise ValueError("Price column empty")
                 end_date = end_series.index[-1]; end_price = end_series.iloc[-1]
                 start_date_target = end_date - pd.DateOffset(years=num_years_cagr)
                 start_series = history_long_df.loc[start_date_target : end_date - pd.Timedelta(days=1), price_col].dropna()
                 if start_series.empty:
                     start_series = history_long_df.loc[:start_date_target, price_col].dropna()
                     if start_series.empty: raise ValueError(f"No data near target start {start_date_target.date()}")
                     start_actual_date = start_series.index[-1]; start_price = start_series.iloc[-1]
                     projection_log.append(f"  ⚠️ Using closest prior date: {start_actual_date.date()}")
                 else: start_actual_date = start_series.index[0]; start_price = start_series.iloc[0]

                 actual_years = (end_date - start_actual_date).days / 365.25
                 projection_log.append(f"  Stock CAGR Points: Start={start_actual_date.date()} ({start_price:.2f}), End={end_date.date()} ({end_price:.2f}), Years={actual_years:.2f}")

                 is_end_valid = pd.notna(end_price) and math.isfinite(end_price)
                 is_start_valid = pd.notna(start_price) and math.isfinite(start_price) and start_price > 1e-9
                 is_period_valid = abs(actual_years - num_years_cagr) < 0.5 and actual_years > 0

                 if is_end_valid and is_start_valid and is_period_valid:
                      stock_cagr = calculate_cagr_modified(start_price, end_price, actual_years)
                      if pd.notna(stock_cagr) and math.isfinite(stock_cagr):
                          calculated_cagrs['Stock'] = stock_cagr; valid_cagr_list.append(stock_cagr)
                          projection_log.append(f"  ✅ Stock CAGR:   {format_percent(stock_cagr)} (Using '{price_col}')")
                      else: projection_log.append(f"  ❌ Stock CAGR Calculation Failed (Result Invalid: {stock_cagr}).")
                 else:
                      reason = f"EndValid={is_end_valid}, StartValid={is_start_valid}, PeriodValid={is_period_valid}"
                      projection_log.append(f"  ❌ Stock CAGR Skipped ({reason}).")
             except Exception as e: projection_log.append(f"  ❌ Stock CAGR Skipped (Error: {e})")
         elif not price_col: pass
         else: projection_log.append(f"  ❌ Stock CAGR Skipped (History index not DatetimeIndex)")
    else: projection_log.append(f"  ❌ Stock CAGR Skipped (Insufficient/invalid history data object)")


    # 4. Calculate Geometric Mean of the valid CAGRs
    projection_log.append("\nCalculating Geometric Mean of Valid CAGRs...")
    if valid_cagr_list:
        # Removed outlier check for simplicity now, rely on bounds later
        gm_growth = geometric_mean_modified(valid_cagr_list)
        if pd.notna(gm_growth) and math.isfinite(gm_growth):
            initial_growth_rate = gm_growth; growth_source = f"GM of {len(valid_cagr_list)} valid CAGR(s)"
            projection_log.append(f"  ▶️ Calculated Geometric Mean = {format_percent(initial_growth_rate)}")
            included_cagrs = [f"{k}:{format_percent(v)}" for k, v in calculated_cagrs.items()]
            projection_log.append(f"     Based on: [{', '.join(included_cagrs)}]")
        else:
            projection_log.append(f"  ⚠️ Geometric Mean calculation failed or result invalid. Falling back to default growth.")
            initial_growth_rate = default_init_growth; growth_source = f"Default ({format_percent(default_init_growth)}) [GM Fail]"
            projection_log.append(f"     CAGRs calculated but GM failed: {valid_cagr_list}")
    else:
        projection_log.append(f"  ⚠️ No valid CAGRs could be calculated. Falling back to default growth.")
        initial_growth_rate = default_init_growth; growth_source = f"Default ({format_percent(default_init_growth)}) [No Valid CAGRs]"

    # 5. Apply Floor (1.1%) and Ceiling (15%) to Initial Growth Rate
    unbounded_rate = initial_growth_rate
    initial_growth_rate = max(initial_growth_rate, MIN_INITIAL_GROWTH_FLOOR); initial_growth_rate = min(initial_growth_rate, MAX_INITIAL_GROWTH_CEILING)
    if abs(initial_growth_rate - unbounded_rate) > 1e-6:
        growth_source += f" (Bounded from {format_percent(unbounded_rate)})"
        projection_log.append(f"  Applied Bounds: Initial Growth Rate = {format_percent(initial_growth_rate)}")
    else:
        projection_log.append(f"  Final Initial Growth Rate = {format_percent(initial_growth_rate)}")

    projection_results['initial_growth_rate'] = initial_growth_rate; projection_results['growth_source'] = growth_source

    # ==============================================================
    # PHASE 4: Determine Terminal Growth & Project FCFF (3 Years)
    # ==============================================================
    projection_log.append("\n--- Phase 4: Determining Terminal Growth & Projecting FCFF ---")

    # --- Terminal Growth Rate Calculation ---
    g_cap_initial_final = initial_growth_rate if pd.notna(initial_growth_rate) and math.isfinite(initial_growth_rate) else TERMINAL_GROWTH_RATE_CAP
    g_cap_wacc = disc_rate * 0.60
    g_cap_rf = final_user_rf * 0.90
    g_cap_config = TERMINAL_GROWTH_RATE_CAP

    projection_log.append(f"\nDetermining Terminal Growth Rate...")
    projection_log.append(f"  Caps Considered: Config={format_percent(g_cap_config)}, WACC*0.60={format_percent(g_cap_wacc)}, Initial(Final)={format_percent(g_cap_initial_final)}, Rf*0.90={format_percent(g_cap_rf)}")

    valid_caps = [cap for cap in [g_cap_config, g_cap_wacc, g_cap_initial_final, g_cap_rf] if np.isfinite(cap)]
    if not valid_caps:
        projection_log.append("  ⚠️ ERROR: No valid caps for terminal growth. Using 0%.")
        terminal_growth_rate = 0.0
    else:
        terminal_growth_rate = min(valid_caps); projection_log.append(f"  ↳ Minimum of Valid Caps: {format_percent(terminal_growth_rate)}")

    original_g_before_floor = terminal_growth_rate
    terminal_growth_rate = max(MIN_GROWTH, terminal_growth_rate)
    if abs(terminal_growth_rate - original_g_before_floor) > 1e-6 : projection_log.append(f"  ↳ Adjusted by Floor ({format_percent(MIN_GROWTH)}): {format_percent(terminal_growth_rate)}")

    if terminal_growth_rate >= disc_rate:
        original_g = terminal_growth_rate;
        terminal_growth_rate = disc_rate - 0.005
        terminal_growth_rate = max(MIN_GROWTH, terminal_growth_rate)
        projection_log.append(f"  ⚠️ Terminal Growth ({format_percent(original_g)}) >= WACC ({format_percent(disc_rate)}). Reduced to {format_percent(terminal_growth_rate)}.")

    projection_log.append(f"  ▶️ Using Final Terminal Growth Rate (g): {format_percent(terminal_growth_rate)}"); projection_results['terminal_growth_rate'] = terminal_growth_rate

    # --- Project FCFFs (3 Years) ---
    projected_fcffs = []; growth_rates_applied = []; pv_fcffs = []; cumulative_pv_fcff = 0.0
    current_fcff_for_proj = fcff_start_basis # Use the basis from Phase 2

    if pd.isna(initial_growth_rate) or not math.isfinite(initial_growth_rate): effective_initial_growth = 0.0
    else: effective_initial_growth = initial_growth_rate
    if pd.isna(terminal_growth_rate) or not math.isfinite(terminal_growth_rate): effective_terminal_growth = 0.0
    else: effective_terminal_growth = terminal_growth_rate

    projection_log.append("\n--- FCFF Projections (3 Years) ---");
    projection_log.append(f"Starting Basis (T=0): {format_currency(current_fcff_for_proj, '', 0, is_indian=is_indian)} (Avg from Method {user_selected_method_idx})")
    projection_log.append(f"{'Year':<5} {'Growth':<10} {'FCFF':<18} {'PV Factor':<10} {'PV FCFF':<18}"); projection_log.append("-" * 65)
    last_valid_fcff = current_fcff_for_proj

    for year in range(1, forecast_yrs + 1): # forecast_yrs is now 3
        if forecast_yrs > 0: decay_factor = max(0, (forecast_yrs - year)) / forecast_yrs
        else: decay_factor = 0
        growth_rate = effective_terminal_growth + (effective_initial_growth - effective_terminal_growth) * decay_factor
        growth_rate = max(MIN_GROWTH, growth_rate)

        growth_rates_applied.append(growth_rate)
        fcff_forecast = last_valid_fcff * (1 + growth_rate)

        if not np.isfinite(fcff_forecast): fcff_forecast = 0.0; projection_log.append(f"⚠️ Year {year} FCFF forecast non-finite. Setting to 0.")
        elif abs(fcff_forecast) > 1e18: fcff_forecast = 0.0; projection_log.append(f"⚠️ Year {year} FCFF forecast too large. Setting to 0.")

        projected_fcffs.append(fcff_forecast)
        last_valid_fcff = fcff_forecast

        pv_factor = 1 / ((1 + disc_rate) ** year); pv_fcff = fcff_forecast * pv_factor
        if not np.isfinite(pv_fcff): pv_fcff = 0.0
        pv_fcffs.append(pv_fcff); cumulative_pv_fcff += pv_fcff
        projection_log.append(f"{year:<5} {format_percent(growth_rate):<10} {format_currency(fcff_forecast, '', 0, is_indian=is_indian):<18} {pv_factor:<10.4f} {format_currency(pv_fcff, '', 0, is_indian=is_indian):<18}")

    projection_results['projected_fcffs'] = projected_fcffs; projection_results['growth_rates'] = growth_rates_applied; projection_results['pv_fcffs'] = pv_fcffs
    projection_results['cumulative_pv_fcff'] = cumulative_pv_fcff if np.isfinite(cumulative_pv_fcff) else 0.0
    projection_log.append("-" * 65); projection_log.append(f"Sum of PV(Forecasted FCFFs): {format_currency(projection_results['cumulative_pv_fcff'], symbol=currency_sym, is_indian=is_indian)}")

    # --- Calculate Terminal Value (Based on Year 3) ---
    projection_log.append("\n--- Terminal Value Calculation ---")
    terminal_value = np.nan; pv_terminal_value = np.nan
    final_forecast_fcff = projected_fcffs[-1] if projected_fcffs else np.nan

    if (disc_rate - effective_terminal_growth) <= 1e-6:
         projection_log.append(f"❌ ERROR: WACC ({format_percent(disc_rate)}) <= Terminal Growth ({format_percent(effective_terminal_growth)}). TV calculation failed."); projection_results['error'] = "WACC <= g. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    elif not np.isfinite(final_forecast_fcff):
         projection_log.append(f"❌ ERROR: Final projected FCFF (Year {forecast_yrs}) is invalid ({final_forecast_fcff}). TV calculation failed."); projection_results['error'] = "Final FCFF invalid. TV Calc failed."
         terminal_value = 0.0; pv_terminal_value = 0.0
    else:
         fcff_n_plus_1 = final_forecast_fcff * (1 + effective_terminal_growth)
         terminal_value = fcff_n_plus_1 / (disc_rate - effective_terminal_growth)

         if not np.isfinite(terminal_value) or terminal_value < 0:
              projection_log.append(f"⚠️ Calculated TV invalid/negative ({format_currency(terminal_value, currency_sym, is_indian=is_indian)}). Setting TV=0.")
              terminal_value = 0.0; pv_terminal_value = 0.0
              if projection_results['error'] is None: projection_results['error'] = "Calculated Terminal Value invalid/negative."
         else:
              pv_factor_tv = 1 / ((1 + disc_rate) ** forecast_yrs); pv_terminal_value = terminal_value * pv_factor_tv
              if not np.isfinite(pv_terminal_value): pv_terminal_value = 0.0; projection_log.append(f"⚠️ PV(TV) calculation failed. Setting PV(TV)=0.")
              projection_log.append(f"FCFF Yr {forecast_yrs}: {format_currency(final_forecast_fcff, currency_sym, is_indian=is_indian)}")
              projection_log.append(f"FCFF Yr {forecast_yrs+1} (Projected): {format_currency(fcff_n_plus_1, currency_sym, is_indian=is_indian)}")
              projection_log.append(f"Terminal Value (at Yr {forecast_yrs}): {format_currency(terminal_value, currency_sym, is_indian=is_indian)}")
              projection_log.append(f"PV Factor TV (Yr {forecast_yrs}): {pv_factor_tv:.4f} -> PV(TV): {format_currency(pv_terminal_value, currency_sym, is_indian=is_indian)}")

    projection_results['terminal_value'] = terminal_value if np.isfinite(terminal_value) else 0.0
    projection_results['pv_terminal_value'] = pv_terminal_value if np.isfinite(pv_terminal_value) else 0.0

    # Display the full log at the end
    display(Markdown("```\n" + "\n".join(projection_log) + "\n```"))
    return projection_results

# --- Execute FCFF calculation and projection ---
# Initialize fcff_results dictionary
fcff_results = {'error': 'Calculation not run yet.', 'initial_growth_rate': np.nan, 'growth_source':'N/A', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'N/A'}

if np.isfinite(discount_rate) and selected_fcff_method is not None:
    try:
        # Use the V21.3 function name
        fcff_results = calculate_and_project_fcff_user_selected_method_v21_3(
            selected_fcff_method, # Pass user's choice
            stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'], stock_data['history_long'],
            wacc_components, discount_rate, FORECAST_YEARS, # FORECAST_YEARS is now 3
            risk_free_rate_input, DEFAULT_GROWTH_RATE,
            currency_symbol, is_indian_stock)
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during FCFF Projection: {e}**"))
         print(traceback.format_exc())
         fcff_results = {'error': f'Runtime error in FCFF calc: {e}', 'initial_growth_rate': np.nan, 'growth_source':'Error', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'Error'}

elif not np.isfinite(discount_rate):
    display(Markdown("❌ **Skipping FCFF Projection due to invalid WACC.**"));
    fcff_results = {'error': 'Invalid WACC.', 'initial_growth_rate': np.nan, 'growth_source':'N/A (WACC Invalid)', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'N/A'}
elif selected_fcff_method is None:
     display(Markdown("❌ **Skipping FCFF Projection because no FCFF method was selected.**"));
     fcff_results = {'error': 'No FCFF method selected.', 'initial_growth_rate': np.nan, 'growth_source':'N/A (No Method)', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'N/A'}


# %% [markdown]
# ## 8. Enterprise Value and Equity Value Calculation
# (Section Unchanged)

# %%
# Valuation Function Definition
def calculate_final_valuation(fcff_data, balance_sheet_df, shares_outstanding_val, currency_sym, wacc_data, is_indian):
    display(Markdown("### Final Valuation Calculation..."))
    valuation = {'enterprise_value': np.nan, 'equity_value': np.nan, 'intrinsic_value_per_share': np.nan, 'error': None}
    valuation_log = []

    if not isinstance(fcff_data, dict):
        error_msg = f"FCFF Data is not a dictionary (type: {type(fcff_data)}). Cannot proceed."
        valuation['error'] = error_msg; valuation_log.append(f"❌ ERROR: {error_msg}")
        display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    fcff_fatal_error = fcff_data.get('error') and "Cannot proceed" in fcff_data.get('error', '')
    if fcff_fatal_error:
         valuation['error'] = f"FCFF Error: {fcff_data.get('error', 'Unknown')}"; valuation_log.append(f"❌ ERROR: Upstream FCFF error prevents valuation: {valuation['error']}"); display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation
    elif fcff_data.get('error'):
         valuation_log.append(f"ℹ️ Proceeding with EV calculation despite FCFF warning: {fcff_data.get('error')}")

    pv_forecast_sum = fcff_data.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_data.get('pv_terminal_value', np.nan)
    if not np.isfinite(pv_forecast_sum): pv_forecast_sum = 0.0; valuation_log.append(f"⚠️ PV(Forecast FCFFs) invalid, using 0.")
    if not np.isfinite(pv_tv): pv_tv = 0.0; valuation_log.append(f"⚠️ PV(Terminal Value) invalid, using 0.")

    enterprise_value = pv_forecast_sum + pv_tv; valuation['enterprise_value'] = enterprise_value
    valuation_log.append(f"PV(Forecast FCFFs): {format_currency(pv_forecast_sum, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"PV(Terminal Value):   {format_currency(pv_tv, currency_sym, is_indian=is_indian)}")
    valuation_log.append(f"↳ Enterprise Value (EV): {format_currency(enterprise_value, currency_sym, is_indian=is_indian)}")

    # Bridge EV to Equity Value
    total_debt_latest = 0.0; cash_and_equivalents = 0.0; minority_interest = 0.0
    debt_source_info = '?';
    if isinstance(wacc_data, dict):
        debt_source_info = wacc_data.get('debt_calc_method', '?');
        total_debt_latest = wacc_data.get('debt_value_used', 0.0)
        if not np.isfinite(total_debt_latest): total_debt_latest = 0.0
    else: valuation_log.append(f"⚠️ WACC data invalid, cannot get debt used in WACC. Assuming 0 for bridge.")

    if balance_sheet_df is not None and not balance_sheet_df.empty:
        latest_bs = balance_sheet_df.iloc[:, 0]
        valuation_log.append(f"\nBridging EV to Equity:")
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        cash_keys = ["Cash And Cash Equivalents", "Cash Cash Equivalents And Short Term Investments", "Cash Financial", "Cash", "Cash And Short Term Investments"];
        cash_and_equivalents = safe_get(latest_bs, cash_keys, "BS (Cash)", default=0)
        if not np.isfinite(cash_and_equivalents): cash_and_equivalents = 0.0
        valuation_log.append(f"  + Cash & Equiv (BS): {format_currency(cash_and_equivalents, currency_sym, is_indian=is_indian)}")
        minority_interest_keys = ["Minority Interest", "Noncontrolling Interest"]
        minority_interest = safe_get(latest_bs, minority_interest_keys, "BS (Minority Int)", default=0)
        if not np.isfinite(minority_interest): minority_interest = 0.0
        if abs(minority_interest) > 1e-6:
             valuation_log.append(f"  - Minority Interest (BS): {format_currency(minority_interest, currency_sym, is_indian=is_indian)}")
    else:
        valuation_log.append("⚠️ Balance Sheet empty or invalid. Cannot fetch Cash/Minority Interest for EV bridge.")
        valuation_log.append(f"  - Total Debt ({debt_source_info}): {format_currency(total_debt_latest, currency_sym, is_indian=is_indian)}")
        valuation_log.append(f"  + Cash & Equiv: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")
        valuation_log.append(f"  - Minority Interest: {format_currency(0.0, currency_sym, is_indian=is_indian)} (Assumed 0)")

    if not all(np.isfinite(v) for v in [enterprise_value, total_debt_latest, cash_and_equivalents, minority_interest]):
         valuation['error'] = "Invalid component(s) for EV->Equity bridge.";
         valuation_log.append(f"❌ ERROR: {valuation['error']} EV={enterprise_value}, Debt={total_debt_latest}, Cash={cash_and_equivalents}, MI={minority_interest}")
         valuation['equity_value'] = np.nan; valuation['intrinsic_value_per_share'] = np.nan
         display(Markdown("```\n" + "\n".join(valuation_log) + "\n```")); return valuation

    equity_value = enterprise_value - total_debt_latest + cash_and_equivalents - minority_interest;
    valuation['equity_value'] = equity_value
    valuation_log.append(f"↳ Equity Value: {format_currency(equity_value, currency_sym, is_indian=is_indian)}")

    intrinsic_value_per_share = np.nan
    if not np.isfinite(equity_value):
        valuation['error'] = f"Equity value invalid ({equity_value})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    elif shares_outstanding_val <= 0:
        valuation['error'] = f"Shares outstanding invalid ({shares_outstanding_val})."; valuation_log.append(f"❌ ERROR: {valuation['error']}")
    else:
        intrinsic_value_per_share = equity_value / shares_outstanding_val;
        valuation_log.append(f"  ÷ Shares Out: {shares_outstanding_val:,.0f}")
        valuation_log.append(f"▶️ Intrinsic Value/Share: {format_currency(intrinsic_value_per_share, currency_sym, is_indian=is_indian)}")

    valuation['intrinsic_value_per_share'] = intrinsic_value_per_share if np.isfinite(intrinsic_value_per_share) else np.nan
    display(Markdown("```\n" + "\n".join(valuation_log) + "\n```"))
    return valuation

# --- Execute Final Valuation ---
final_valuation = {'error': 'Calculation not run yet.', 'intrinsic_value_per_share': np.nan}
fcff_error_is_fatal = isinstance(fcff_results, dict) and fcff_results.get('error') and "Cannot proceed" in fcff_results.get('error', '')

if isinstance(fcff_results, dict) and isinstance(wacc_components, dict) and not fcff_error_is_fatal:
    try:
        final_valuation = calculate_final_valuation(
            fcff_results, stock_data['balance_sheet'], shares_outstanding,
            currency_symbol, wacc_components, is_indian_stock
        )
    except Exception as e:
        display(Markdown(f"❌ **CRITICAL ERROR during Final Valuation: {e}**")); print(traceback.format_exc())
        final_valuation = {'error': f'Runtime error in final valuation: {e}', 'intrinsic_value_per_share': np.nan}
elif isinstance(fcff_results, dict) and fcff_error_is_fatal:
     display(Markdown(f"❌ **Skipping Final Valuation due to fatal FCFF error:** {fcff_results.get('error')}"));
     final_valuation = {'error': fcff_results.get('error', 'Upstream fatal FCFF error.'), 'intrinsic_value_per_share': np.nan}
else:
     display(Markdown("❌ **Skipping Final Valuation due to critical upstream errors (invalid WACC/FCFF data structures).**"));
     final_valuation = {'error': 'FCFF or WACC results dictionary not created/valid.', 'intrinsic_value_per_share': np.nan}


# %% [markdown]
# ## 9. Financial Ratio Analysis
# (Section Unchanged - Plotting Fixes Already Applied)

# %%
# Ratio Function Definition (V21.2 Plot Fixes Retained)
def analyze_ratios(stock_info_dict, financials_df, balance_sheet_df, cashflow_df,
                   enterprise_val, current_price_val, shares_outstanding_val,
                   currency_sym, user_benchmark_dict, # Benchmarks stored as entered (e.g., 15.0 for 15%)
                   benchmark_source_dict, # Dict tracking source per ratio
                   wacc_data, is_indian): # Pass WACC data for consistent debt and flag

    """ Calculates ratios, compares to user/fallback benchmarks, uses WACC-consistent debt. Plots fixed. """
    display(Markdown("### Financial Ratio Analysis..."))
    ratios = {} # Calculated ratios {RatioName: value (decimal form)}
    ratio_log = []
    ratio_data_for_table = [] # For HTML table display
    ratio_plotly_data = {'Ratio': [], 'Company': [], 'Benchmark': [], 'Benchmark Source': [], 'Status': []} # For Plotly chart
    company_nm = stock_info_dict.get("shortName", ticker_symbol)

    # --- Pre-checks ---
    valid_price = current_price_val is not None and np.isfinite(current_price_val) and current_price_val > 0
    valid_shares = shares_outstanding_val is not None and np.isfinite(shares_outstanding_val) and shares_outstanding_val > 0
    valid_enterprise_val = enterprise_val is not None and np.isfinite(enterprise_val)

    if not valid_price: ratio_log.append("⚠️ Current Price invalid/unavailable. Price-based ratios (P/E, P/B) will be N/A.")
    if not valid_shares: ratio_log.append("⚠️ Shares Outstanding invalid/unavailable. Per-share ratios (EPS, BVPS, P/E, P/B) will be N/A.")
    if not valid_enterprise_val: ratio_log.append("⚠️ Enterprise Value invalid/unavailable. EV/EBITDA will be N/A.")


    essential_statements_missing = False
    if financials_df is None or financials_df.empty:
         ratio_log.append("❌ ERROR: Income Statement empty. Most ratios cannot be calculated."); essential_statements_missing = True
    if balance_sheet_df is None or balance_sheet_df.empty:
         ratio_log.append("❌ ERROR: Balance Sheet empty. Many ratios cannot be calculated."); essential_statements_missing = True
    if cashflow_df is None or cashflow_df.empty:
         ratio_log.append("⚠️ Warning: Cash Flow statement empty. EBITDA calculation might be inaccurate (assuming 0 Depreciation).")

    if essential_statements_missing:
         display(Markdown("```\n" + "\n".join(ratio_log) + "\n```")); return {}, []

    latest_fin = financials_df.iloc[:, 0] if not financials_df.empty else pd.Series(dtype=float)
    latest_bs = balance_sheet_df.iloc[:, 0] if not balance_sheet_df.empty else pd.Series(dtype=float)
    latest_cf = cashflow_df.iloc[:, 0] if not cashflow_df.empty else pd.Series(dtype=float)

    # --- Calculate Ratio Components ---
    net_income = safe_get(latest_fin, ["Net Income Continuous Operations", "Net Income", "Net Income Common Stockholders", "Net Income Applicable To Common Shares"], default=np.nan)
    revenue = safe_get(latest_fin, ["Total Revenue", "Revenue", "Net Revenue"], default=np.nan)
    ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], default=np.nan)
    depreciation = safe_get(latest_cf, ["Depreciation And Amortization", "Depreciation", "Depreciation Amortization Depletion"], "Cash Flow (Dep)", default=np.nan)
    if not np.isfinite(depreciation):
        dep_is = safe_get(latest_fin, ["Depreciation And Amortization", "Depreciation Income Statement"], "Income Stmt (Dep Fallback)", default=np.nan)
        if np.isfinite(dep_is): depreciation = dep_is; ratio_log.append("ℹ️ Used Depreciation from Income Statement (CF unavailable/invalid).")
        else: depreciation = 0.0; ratio_log.append("⚠️ Depreciation not found in CF or IS. Assuming 0 for EBITDA calc.")
    else: depreciation = float(depreciation)

    ebitda = np.nan
    if np.isfinite(ebit) and np.isfinite(depreciation): ebitda = ebit + depreciation
    elif np.isfinite(ebit): ebitda = ebit; ratio_log.append("ℹ️ Calculating EBITDA using EBIT only (Depreciation unavailable/zero).")
    else: ratio_log.append("⚠️ EBIT invalid. Cannot calculate EBITDA.")

    total_equity = safe_get(latest_bs, ["Total Stockholder Equity", "Stockholders Equity", "Total Equity Gross Minority Interest", "Common Stock Equity"], "BS (Equity)", default=np.nan)
    total_assets = safe_get(latest_bs, ["Total Assets"], "BS (Assets)", default=np.nan)
    current_assets = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], "BS (Curr Assets)", default=np.nan)
    current_liab = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], "BS (Curr Liab)", default=np.nan)
    inventory = safe_get(latest_bs, ["Inventory"], "BS (Inventory)", default=np.nan)
    if not np.isfinite(inventory): inventory = 0.0; ratio_log.append("ℹ️ Inventory not found on BS. Assuming 0 for Quick Ratio calc.")
    else: inventory = float(inventory)

    total_debt_ratio = 0.0; debt_source_info = '?'
    if isinstance(wacc_data, dict):
        total_debt_ratio = wacc_data.get('debt_value_used', 0.0)
        debt_source_info = wacc_data.get('debt_calc_method', '?');
        if not np.isfinite(total_debt_ratio): total_debt_ratio = 0.0
    else: ratio_log.append(f"⚠️ WACC data invalid, cannot get consistent debt. Assuming 0 for ratio calcs."); total_debt_ratio = 0.0
    ratio_log.append(f"Using Debt = {format_currency(total_debt_ratio, currency_sym, is_indian=is_indian)} (Source: WACC - {debt_source_info}) for relevant ratios.")

    eps = (net_income / shares_outstanding_val) if valid_shares and np.isfinite(net_income) and net_income is not None else np.nan
    bvps = (total_equity / shares_outstanding_val) if valid_shares and np.isfinite(total_equity) and total_equity is not None else np.nan

    capital_employed = np.nan; ce_source = "N/A"
    if np.isfinite(total_assets) and np.isfinite(current_liab):
        capital_employed_def1 = total_assets - current_liab
        if np.isfinite(capital_employed_def1) and capital_employed_def1 > 1.0: capital_employed = capital_employed_def1; ce_source = "Total Assets - Current Liab"
    if not np.isfinite(capital_employed):
        if np.isfinite(total_equity) and np.isfinite(total_debt_ratio):
             capital_employed_def2 = total_equity + total_debt_ratio
             if np.isfinite(capital_employed_def2) and capital_employed_def2 > 1.0: capital_employed = capital_employed_def2; ce_source = "Total Equity + Total Debt"
    if not np.isfinite(capital_employed): ce_source = "Invalid/Unavailable"; ratio_log.append("⚠️ Capital Employed could not be reliably calculated (needed for ROCE). Check BS data.")

    # --- Calculate Ratios (Result is ratio value, e.g., 0.15 for 15% ROE) ---
    def safe_division(numerator, denominator):
        num_val = float(numerator) if np.isfinite(numerator) else np.nan
        den_val = float(denominator) if np.isfinite(denominator) else np.nan
        if not np.isfinite(num_val) or not np.isfinite(den_val): return np.nan
        if abs(den_val) < 1e-9: return np.nan
        try: return num_val / den_val
        except: return np.nan

    ratios['P/E'] = safe_division(current_price_val, eps) if valid_price else np.nan
    ratios['P/B'] = safe_division(current_price_val, bvps) if valid_price else np.nan
    ratios['EV/EBITDA'] = safe_division(enterprise_val, ebitda) if valid_enterprise_val else np.nan
    ratios['ROE'] = safe_division(net_income, total_equity)
    ratios['ROA'] = safe_division(net_income, total_assets)
    ratios['ROCE'] = safe_division(ebit, capital_employed)
    ratios['Asset Turnover'] = safe_division(revenue, total_assets)
    ratios['Current Ratio'] = safe_division(current_assets, current_liab)
    quick_assets = (current_assets - inventory) if np.isfinite(current_assets) and np.isfinite(inventory) else np.nan
    ratios['Quick Ratio'] = safe_division(quick_assets, current_liab)

    # --- Compare with User/Fallback Benchmarks & Prepare Data Tables ---
    display(Markdown("*(Comparing ratios to user/fallback benchmarks...)*"))
    for key in REQUIRED_BENCHMARK_RATIOS:
        company_value = ratios.get(key); is_percent_type = key in ['ROE', 'ROA', 'ROCE']
        value_fmt = "N/A"
        if company_value is None or pd.isna(company_value): value_fmt = "N/A"
        elif not math.isfinite(company_value): value_fmt = "Inf" if company_value > 0 else "-Inf"
        else:
            if is_percent_type: value_fmt = format_percent(company_value, 1)
            elif key == 'Asset Turnover': value_fmt = f"{company_value:.2f}"
            else: value_fmt = format_currency(company_value, '', 2, is_indian=False)

        user_benchmark_val = user_benchmark_dict.get(key, None)
        fallback_benchmark_val = DEFAULT_BENCHMARK_RATIOS.get(key, None)
        benchmark_source = benchmark_source_dict.get(key, "?"); final_benchmark_val = None
        if benchmark_source == 'User Input' and user_benchmark_val is not None and np.isfinite(user_benchmark_val): final_benchmark_val = user_benchmark_val
        elif benchmark_source == 'Fallback Default' and fallback_benchmark_val is not None and np.isfinite(fallback_benchmark_val): final_benchmark_val = fallback_benchmark_val
        else: benchmark_source = "Not Available"

        bench_fmt = "N/A"
        if final_benchmark_val is not None:
             if is_percent_type: bench_fmt = f"{final_benchmark_val:.1f}%"
             elif key == 'Asset Turnover': bench_fmt = f"{final_benchmark_val:.2f}"
             else: bench_fmt = f"{final_benchmark_val:.2f}"
             bench_fmt += f" <i style='font-size:0.8em; color:grey'>({benchmark_source.replace('Fallback ','')})</i>"
        else: bench_fmt = "<i style='color:grey'>Not Provided</i>"

        status = "<i style='color:grey'>N/A</i>"; status_color = "grey"
        if np.isfinite(company_value) and final_benchmark_val is not None and np.isfinite(final_benchmark_val):
            numeric_value_comp = company_value * 100 if is_percent_type else company_value
            numeric_benchmark_comp = final_benchmark_val # Benchmark already stored as % value if needed
            if key == 'Asset Turnover': numeric_benchmark_comp = final_benchmark_val # Keep AT as decimal

            good_color = '#2ecc71'; bad_color = '#e74c3c'; neutral_color = '#f39c12'; grey_color='#7f8c8d'
            if abs(numeric_benchmark_comp) > 1e-9: ratio_diff = (numeric_value_comp - numeric_benchmark_comp) / abs(numeric_benchmark_comp)
            else: ratio_diff = float('inf') if numeric_value_comp > 0 else float('-inf') if numeric_value_comp < 0 else 0.0

            if key in ['ROE', 'ROA', 'ROCE', 'Asset Turnover', 'Current Ratio', 'Quick Ratio']:
                 if ratio_diff > 0.20: status_color = good_color; status = f"⬆️ Strong"
                 elif ratio_diff < -0.20: status_color = bad_color; status = f"⬇️ Weak"
                 else: status_color = grey_color; status = f"~ Average"
            elif key in ['P/E', 'P/B', 'EV/EBITDA']:
                 if ratio_diff < -0.20: status_color = good_color; status = f"✅ Low (vs Bench)"
                 elif ratio_diff > 0.50: status_color = bad_color; status = f"‼️ High (vs Bench)"
                 elif ratio_diff > 0.20: status_color = neutral_color; status = f"⚠️ Elevated (vs Bench)"
                 else: status_color = grey_color; status = f"~ Average (vs Bench)"
            status = f"<span style='color:{status_color}'>{status}</span>"

            plot_value = numeric_value_comp
            plot_benchmark = numeric_benchmark_comp
            ratio_plotly_data['Ratio'].append(key);
            ratio_plotly_data['Company'].append(plot_value);
            ratio_plotly_data['Benchmark'].append(plot_benchmark);
            ratio_plotly_data['Benchmark Source'].append(benchmark_source);
            ratio_plotly_data['Status'].append(status)
        elif not np.isfinite(company_value): status = "<i style='color:grey'>Company N/A</i>"
        elif final_benchmark_val is None: status = "<i style='color:grey'>No Benchmark</i>"
        ratio_data_for_table.append([f"<b>{key}</b>", value_fmt, bench_fmt, status])

    # --- Display Logs & Table ---
    if ratio_log: display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
    if ratio_data_for_table:
        header = ["Ratio", f"{company_nm} Value", "Benchmark (Source)", "Status"]; html_table = "<p>(*Status vs User/Default benchmarks...)*</p>"; html_table += "<table border='1' style='border-collapse:collapse; width:98%; margin-left:auto; margin-right:auto; font-size: 0.9em;'>"; html_table += "<thead style='background-color:#e9ecef; color: #495057;'><tr>" + "".join(f"<th style='padding: 8px; text-align:center;'>{h}</th>" for h in header) + "</tr></thead>"; html_table += "<tbody>"
        for i, row in enumerate(ratio_data_for_table): bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff"; html_table += f"<tr style='background-color:{bg_color};'>"; html_table += f"<td style='padding: 6px; text-align:left;'>{row[0]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[1]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[2]}</td>"; html_table += f"<td style='padding: 6px; text-align:center;'>{row[3]}</td>"; html_table += "</tr>"
        html_table += "</tbody></table><br>"; display(HTML(html_table))
    else: display(Markdown("*No ratio data available for table display.*"))

    # --- Display Ratio Comparison Chart (Plotly Bullet - V21.2 Fixes Retained) ---
    if ratio_plotly_data['Ratio']:
       num_ratios = len(ratio_plotly_data['Ratio']); fig_ratios = go.Figure()
       good_color = '#90EE90'; bad_color = '#F08080'; avg_color = '#D3D3D3'

       def get_range_colors(k):
           if k in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio']: return [bad_color, avg_color, good_color]
           elif k in ['P/E','P/B','EV/EBITDA']: return [good_color, avg_color, bad_color]
           else: return ['lightgrey','darkgrey','grey']

       for i in range(num_ratios):
           ratio_name = ratio_plotly_data['Ratio'][i];
           company_val = ratio_plotly_data['Company'][i]; benchmark_val = ratio_plotly_data['Benchmark'][i];
           bench_src_plot = ratio_plotly_data['Benchmark Source'][i]
           is_percent = ratio_name in ['ROE', 'ROA', 'ROCE']

           if not np.isfinite(company_val) or not np.isfinite(benchmark_val):
               print(f"Skipping plot for {ratio_name}: Invalid Value ({company_val}) or Bench ({benchmark_val}).")
               continue

           range_colors = get_range_colors(ratio_name);
           range_delta = 15.0 if is_percent else max(1.0, abs(benchmark_val * 0.2))
           if ratio_name == 'Asset Turnover': range_delta = 0.2
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: range_delta = 0.5
           elif ratio_name in ['P/E', 'P/B', 'EV/EBITDA']: range_delta = max(2.0, abs(benchmark_val*0.2))

           range1_limit = benchmark_val - range_delta
           range2_limit = benchmark_val + range_delta

           abs_max_val = max(abs(company_val), abs(benchmark_val), 1)
           plot_max_val = max(company_val, benchmark_val) + range_delta * 1.5
           plot_min_val = min(company_val, benchmark_val) - range_delta * 1.5

           if ratio_name in ['P/E', 'EV/EBITDA']: plot_max_val = min(plot_max_val, 100); plot_min_val = max(plot_min_val, -20)
           elif ratio_name == 'P/B': plot_max_val = min(plot_max_val, 20); plot_min_val = max(plot_min_val, -5)
           elif ratio_name in ['Current Ratio', 'Quick Ratio']: plot_max_val = min(plot_max_val, 5); plot_min_val = max(plot_min_val, 0)
           elif is_percent: plot_max_val = min(plot_max_val, 150); plot_min_val = max(plot_min_val,-100)
           elif ratio_name == 'Asset Turnover': plot_max_val = min(plot_max_val, 3); plot_min_val = max(plot_min_val, -0.5)

           if plot_min_val >= plot_max_val: plot_max_val = plot_min_val + range_delta

           steps = [];
           step_boundaries = sorted([plot_min_val, range1_limit, range2_limit, plot_max_val])
           step_colors = range_colors
           for j in range(len(step_colors)):
               start_range = step_boundaries[j]; end_range = step_boundaries[j+1]
               if start_range < end_range: steps.append({'range': [start_range, end_range], 'color': step_colors[j]})

           if abs(benchmark_val) < 1e-9:
                range1_limit_zero = -range_delta / 2; range2_limit_zero = range_delta / 2
                step_boundaries_zero = sorted([plot_min_val, range1_limit_zero, range2_limit_zero, plot_max_val])
                steps = []
                for j in range(len(step_colors)):
                    start_range = step_boundaries_zero[j]; end_range = step_boundaries_zero[j+1]
                    if start_range < end_range: steps.append({'range': [start_range, end_range], 'color': step_colors[j]})

           bench_fmt_title = f"{benchmark_val:.1f}%" if is_percent else f"{benchmark_val:.2f}";
           title_text = f"<b>{ratio_name}</b><br><i style='font-size:0.8em'>(Bench: {bench_fmt_title} [{bench_src_plot.replace('Fallback ','')}])</i>"
           lower_is_better = ratio_name in ['P/E', 'P/B', 'EV/EBITDA']
           delta_decreasing_color = '#90EE90' if lower_is_better else '#F08080';
           delta_increasing_color = '#F08080' if lower_is_better else '#90EE90';

           fig_ratios.add_trace(go.Indicator(
                mode = "number+gauge+delta", value = company_val,
                number={'suffix': "%" if is_percent else "", 'valueformat': ".1f" if is_percent else ".2f", 'font':{'size':16}},
                delta = {'reference': benchmark_val, 'relative': False, 'valueformat':".1f" if is_percent else ".2f",
                         'decreasing': {'color': delta_decreasing_color}, 'increasing': {'color': delta_increasing_color}},
                domain = {'row': i, 'column': 0}, title = {'text' : title_text, 'font':{'size':11}},
                gauge = { 'shape': "bullet", 'axis': {'range': [plot_min_val, plot_max_val], 'tickformat':'.1f'},
                         'threshold': { 'line': {'color': "black", 'width': 2}, 'thickness': 0.8, 'value': benchmark_val },
                         'steps': steps, 'bar': {'color': "#2c3e50", 'thickness': 0.5} }
            ))
       fig_ratios.update_layout(height = max(300, num_ratios * 90), grid = {'rows': num_ratios, 'columns': 1, 'pattern': "independent"},
           title_text='Ratio Performance vs. Benchmarks', title_x=0.5, margin=dict(l=180, r=40, t=60, b=30))
       fig_ratios.show()
    else: display(Markdown("*No valid ratios/benchmarks available for comparison chart.*"))

    return ratios, ratio_data_for_table

# --- Execute Ratio Analysis ---
ratios_calculated = {}; ratio_table = []
valuation_ready = isinstance(final_valuation, dict) and not (final_valuation.get('error') and "Cannot proceed" in final_valuation.get('error',''))
wacc_ready = isinstance(wacc_components, dict)

if valuation_ready and wacc_ready:
    try:
        ev_for_ratios = final_valuation.get('enterprise_value', np.nan)
        if not np.isfinite(ev_for_ratios): print("WARNING: EV for ratios is invalid. EV/EBITDA will fail.")
        ratios_calculated, ratio_table = analyze_ratios(
            stock_data['info'], stock_data['financials'], stock_data['balance_sheet'], stock_data['cashflow'],
            ev_for_ratios, current_price, shares_outstanding, currency_symbol,
            user_benchmarks, benchmark_sources, wacc_components, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Ratio Analysis: {e}**")); print(traceback.format_exc())
         ratios_calculated = {'error': f'Runtime error in ratio analysis: {e}'}; ratio_table = []
else:
    error_reason = final_valuation.get('error', 'Unknown Valuation Error') if not valuation_ready else 'WACC Components Invalid'
    display(Markdown(f"❌ **Skipping Ratio Analysis due to prior errors:** {error_reason}"));
    ratios_calculated = {'error': f'Skipped due to: {error_reason}'}; ratio_table = []


# %% [markdown]
# ## 10. Summary & Visualization Dashboard
# (Section Unchanged)

# %%
# Dashboard Function Definition
def display_summary_dashboard(company_nm, ticker_sym, current_pr, intrinsic_val_per_sh,
                              wacc_val, term_growth, init_growth, growth_src,
                              final_val_dict, hist_df, fcff_res, currency_sym, info_dict, input_src_dict,
                              is_indian):

    display(Markdown(f"## Valuation Summary Dashboard: {company_nm} ({ticker_sym})"))
    source_items = []
    for k, v in input_src_dict.items():
        if isinstance(v, str): source_items.append(f"{k}:`{v.split('(')[0].strip()}`")
    source_summary = " | ".join(source_items)

    growth_src_display = growth_src if isinstance(growth_src, str) else 'N/A'
    fcff_basis_method_disp = fcff_res.get('fcff_basis_method', 'N/A') if isinstance(fcff_res, dict) else 'N/A'
    display(Markdown(f"**Input Sources:** {source_summary} | Growth:`{growth_src_display}` | FCFF Basis:`{fcff_basis_method_disp}`"))

    valid_intrinsic = intrinsic_val_per_sh is not None and np.isfinite(intrinsic_val_per_sh)
    valid_current_pr = current_pr is not None and np.isfinite(current_pr) and current_pr > 0
    valid_wacc = wacc_val is not None and np.isfinite(wacc_val)
    valid_term_g = term_growth is not None and np.isfinite(term_growth)
    valid_init_g = init_growth is not None and np.isfinite(init_growth)

    # --- Indicator Subplots ---
    fig_indicators = make_subplots(rows=1, cols=3, specs=[[{'type': 'indicator'}, {'type': 'indicator'}, {'type': 'indicator'}]],
                                 subplot_titles=("Price vs. Intrinsic", "WACC", "Growth Rates"), horizontal_spacing=0.1, column_widths=[0.4, 0.3, 0.3])

    # Indicator 1: Price vs Intrinsic
    intrinsic_val_num = intrinsic_val_per_sh if valid_intrinsic else np.nan
    intrinsic_val_fmt = format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian) if valid_intrinsic else "N/A"
    current_pr_num = current_pr if valid_current_pr else 0.0; upside_fmt = ""; delta_config = None
    comparison_text = f"Intrinsic: {intrinsic_val_fmt}"

    if valid_intrinsic and valid_current_pr :
        if intrinsic_val_num > 1e-9:
             upside_pct = (intrinsic_val_num / current_pr_num) - 1; upside_fmt = f" ({upside_pct:+.1%})"
             comparison_text += upside_fmt
             delta_config = {'reference': intrinsic_val_num, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym,
                             'decreasing': {'color': '#2ecc71', 'symbol': '▲ Upside'}, 'increasing': {'color': '#e74c3c', 'symbol': '▼ Downside'}}
        else:
             comparison_text += " (≤ 0)"
             delta_config = {'reference': 0, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym + " " if is_indian else currency_sym,
                             'decreasing': {'color': 'orange', 'symbol': ''}, 'increasing': {'color': '#e74c3c', 'symbol': '▼ Price > 0'}}
    elif valid_intrinsic: comparison_text += " (Current N/A)"
    else: comparison_text = f"Intrinsic: N/A (Calculation Error)"

    fig_indicators.add_trace(go.Indicator(mode = "number+delta" if delta_config and valid_current_pr else "number",
        value = current_pr_num if valid_current_pr else 0,
        number={'prefix': currency_sym + " " if is_indian else currency_sym, 'valueformat': '.2f', 'font':{'size':24}},
        title = {"text": f"<span style='font-size:0.8em'>{comparison_text}</span>"},
        delta = delta_config if valid_current_pr else None,
        domain = {'row': 0, 'column': 0}), row=1, col=1)

    # Indicator 2: WACC Gauge
    wacc_val_num = wacc_val * 100 if valid_wacc else None
    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = wacc_val_num,
        number = {'suffix': "%", 'valueformat':'.1f' if valid_wacc else '', 'font':{'size':20}},
        gauge = {'axis': {'range': [MIN_WACC*100, MAX_WACC*100]}, 'bar': {'color': "darkblue" if valid_wacc else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [ {'range': [MIN_WACC*100, 9], 'color': "#d5f5e3"}, {'range': [9, 14], 'color': "#fdebd0"}, {'range': [14, MAX_WACC*100], 'color': "#fadbd8"}]} if valid_wacc else None,
        title = {"text": "<span style='font-size:0.8em'>WACC</span>"}, domain = {'row': 0, 'column': 1}), row=1, col=2)

    # Indicator 3: Growth Gauge
    term_growth_num = term_growth * 100 if valid_term_g else None; init_growth_fmt = format_percent(init_growth, 1) if valid_init_g else "N/A"
    growth_title_text = f"<span style='font-size:0.8em'>Terminal Growth (g)<br>(Initial: {init_growth_fmt})</span>"
    gauge_max_range = max(TERMINAL_GROWTH_RATE_CAP * 100 + 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) + 2)
    gauge_min_range = min(MIN_GROWTH * 100 - 1, (term_growth_num if valid_term_g and term_growth_num is not None else 0) - 2)

    fig_indicators.add_trace(go.Indicator(mode = "gauge+number", value = term_growth_num,
         number = {'suffix': "%", 'valueformat':'.1f' if valid_term_g else '', 'font':{'size':20}},
         gauge = {'axis': {'range': [gauge_min_range, gauge_max_range], 'tickwidth': 1}, 'bar': {'color': "darkgrey" if valid_term_g else "lightgrey", 'thickness': 0.4}, 'bgcolor': "white", 'borderwidth': 1, 'bordercolor': "#cccccc",
                 'steps' : [{'range': [gauge_min_range, 2.0], 'color': "#fadbd8"}, {'range': [2.0, 4.5], 'color': "#fdebd0"}, {'range': [4.5, gauge_max_range], 'color': "#d5f5e3"}],
                 'threshold' : {'line': {'color': "black", 'width': 3}, 'thickness': 0.75, 'value': term_growth_num } if valid_term_g else None } if valid_term_g else None,
         title = {"text": growth_title_text}, domain = {'row': 0, 'column': 2}), row=1, col=3)

    fig_indicators.update_layout(height=220, margin=dict(l=30, r=30, t=50, b=20)); fig_indicators.show()

    # --- Historical Price Chart ---
    if hist_df is not None and not hist_df.empty:
        price_col_hist = None
        if 'Adj Close' in hist_df.columns and hist_df['Adj Close'].notna().any(): price_col_hist = 'Adj Close'
        elif 'Close' in hist_df.columns and hist_df['Close'].notna().any(): price_col_hist = 'Close'

        if price_col_hist:
             fig_hist = px.line(hist_df, x=hist_df.index, y=price_col_hist, title=f'{company_nm} Price History ({price_col_hist}) vs. DCF');
             fig_hist.update_layout(xaxis_title='Date', yaxis_title=f'Price ({currency_sym})', height=400, legend_title_text='Legend'); fig_hist.update_xaxes(rangeslider_visible=False)
             if valid_intrinsic: fig_hist.add_hline(y=intrinsic_val_num, line_dash="dash", line_color="#e74c3c", opacity=0.8, annotation_text=f"DCF Intrinsic: {format_currency(intrinsic_val_num, currency_sym, is_indian=is_indian)}", annotation_position="bottom right", annotation_font_color="#e74c3c")
             if valid_current_pr: fig_hist.add_hline(y=current_pr_num, line_dash="dot", line_color="#3498db", opacity=0.8, annotation_text=f"Current: {format_currency(current_pr_num, currency_sym, is_indian=is_indian)}", annotation_position="top left", annotation_font_color="#3498db")
             fig_hist.show()
        else: display(Markdown("*Historical price data ('Adj Close' or 'Close') unavailable for chart.*"))
    else: display(Markdown("*Historical price data object unavailable.*"))

    # --- DCF EV Breakdown Pie Chart ---
    if isinstance(fcff_res, dict):
        pv_forecast_sum = fcff_res.get('cumulative_pv_fcff', np.nan); pv_tv = fcff_res.get('pv_terminal_value', np.nan)
        valid_pvs = np.isfinite([pv_forecast_sum, pv_tv]);
        total_ev = np.nansum([pv_forecast_sum, pv_tv]) if any(valid_pvs) else np.nan
        fcff_fatal_error = fcff_res.get('error') and "Cannot proceed" in fcff_res.get('error', '')

        if np.isfinite(total_ev) and not fcff_fatal_error:
            labels_pie = []; values_pie = []; pull_pie = []; colors = ['#4e79a7','#f28e2b']
            if abs(pv_forecast_sum) > 1e-6 : labels_pie.append(f'PV({FORECAST_YEARS}yr FCFFs)'); values_pie.append(abs(pv_forecast_sum)); pull_pie.append(0)
            if abs(pv_tv) > 1e-6 : labels_pie.append('PV(TV)'); values_pie.append(abs(pv_tv)); pull_pie.append(0.05 if len(labels_pie) > 1 else 0)

            if labels_pie:
                hover_format = f"<b>%{{label}}</b><br>Value: {currency_sym} %{{value:,.0f}}<br>%{{percent}}<extra></extra>"
                fig_dcf_pie = go.Figure(data=[go.Pie(labels=labels_pie, values=values_pie, hole=.4, pull=pull_pie, marker_colors=colors[:len(labels_pie)], hovertemplate=hover_format, textinfo='percent+label', insidetextorientation='radial', sort=False)])
                fig_dcf_pie.update_layout( title_text=f"DCF EV Breakdown (Total EV: {format_currency(total_ev, currency_sym, is_indian=is_indian)})", title_x=0.5, legend_title_text='Components (Abs Val)', legend=dict(orientation="h", yanchor="bottom", y=-0.1, xanchor="center", x=0.5), height=350, margin=dict(t=60, b=50, l=20, r=20) )
                fig_dcf_pie.show()
            else: display(Markdown(f"*DCF components zero/negligible or negative. Cannot generate EV Breakdown chart.*"))
        elif fcff_fatal_error: display(Markdown(f"*Cannot generate EV Breakdown chart due to fatal FCFF projection error: {fcff_res.get('error')}*"))
        else: display(Markdown(f"*EV calculation failed or invalid ({format_currency(total_ev, currency_sym, is_indian=is_indian)}). Cannot generate EV Breakdown chart.*"))
    else: display(Markdown(f"*FCFF results invalid. Cannot generate EV Breakdown chart.*"))

    # --- FCFF Projection Bar Chart ---
    if isinstance(fcff_res, dict) and fcff_res.get('projected_fcffs') and not fcff_fatal_error:
        years = [f'Y{i+1}' for i in range(len(fcff_res['projected_fcffs']))];
        pv_list_clean = [p if np.isfinite(p) else 0 for p in fcff_res.get('pv_fcffs', [])];
        fcff_list_clean = [f if np.isfinite(f) else 0 for f in fcff_res.get('projected_fcffs', [])]

        if len(pv_list_clean) == len(years) and len(fcff_list_clean) == len(years) and len(years) > 0:
            df_fcff_plot = pd.DataFrame({'Year': years, 'Forecasted FCFF': fcff_list_clean, 'PV of FCFF': pv_list_clean});
            df_fcff_plot['Forecasted FCFF'] = pd.to_numeric(df_fcff_plot['Forecasted FCFF']);
            df_fcff_plot['PV of FCFF'] = pd.to_numeric(df_fcff_plot['PV of FCFF'])
            df_fcff_melt = df_fcff_plot.melt(id_vars='Year', var_name='Cash Flow Type', value_name='Value')
            fig_fcff_bar = px.bar(df_fcff_melt, x='Year', y='Value', color='Cash Flow Type', title='FCFF Forecast & Present Values (3 Years)', labels={'Value':f'Value ({currency_sym})'}, barmode='group', height=350, color_discrete_map={'Forecasted FCFF':'#5dade2', 'PV of FCFF':'#af7ac5'})
            hover_bar_format = f"<b>%{{data.name}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            fig_fcff_bar.update_traces(hovertemplate=hover_bar_format); fig_fcff_bar.update_layout(margin=dict(t=60, b=30, l=50, r=30)); fig_fcff_bar.show()
        else: display(Markdown("*FCFF projection data mismatch/invalid or empty. Cannot plot.*"))
    elif fcff_fatal_error: display(Markdown(f"*Cannot generate FCFF Projection chart due to fatal error: {fcff_res.get('error')}*"))
    else: display(Markdown("*FCFF projection data unavailable or invalid.*"))

    # --- Valuation Conclusion ---
    display(Markdown("### Valuation Conclusion"));
    if isinstance(final_val_dict, dict):
        intrinsic_val = final_val_dict.get('intrinsic_value_per_share', np.nan)
        val_error = final_val_dict.get('error', None)
        fcff_error = fcff_res.get('error', None) if isinstance(fcff_res, dict) else "FCFF Data Invalid"
        conclusion = ""

        if val_error: conclusion = f"<span style='color:red'>❌ Valuation Error: {val_error}</span>"
        elif fcff_fatal_error: conclusion = f"<span style='color:red'>❌ Valuation Failed: {fcff_error}</span>"
        elif not valid_intrinsic: conclusion = f"<span style='color:red'>❌ Valuation Error: Intrinsic Value calculation failed (result is {intrinsic_val}).</span>"
        elif valid_intrinsic and intrinsic_val <= 0: conclusion = f"<span style='color:orange'>⚠️ Negative/Zero DCF Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}.</span>"
        elif valid_intrinsic and not valid_current_pr: conclusion = f"✅ DCF Intrinsic Value: {format_currency(intrinsic_val, currency_sym, is_indian=is_indian)}. Cannot compare (Current Price N/A)."
        elif valid_intrinsic and valid_current_pr :
             diff_pct = (intrinsic_val / current_pr_num) - 1
             current_fmt = format_currency(current_pr_num, currency_sym, is_indian=is_indian)
             intrinsic_fmt = format_currency(intrinsic_val, currency_sym, is_indian=is_indian)
             if abs(diff_pct) < 0.10: conclusion = f"<span style='color:grey'>~ Fairly Valued:</span> Price ({current_fmt}) vs DCF ({intrinsic_fmt}). Diff: {diff_pct:+.1%}"
             elif diff_pct < 0: overval_pct = (current_pr_num / intrinsic_val) - 1 if intrinsic_val != 0 else float('inf'); conclusion = f"<span style='color:#e74c3c'>🔻 Potentially Overvalued:</span> Price ({current_fmt}) is **{overval_pct:+.1%}** above DCF ({intrinsic_fmt})."
             else: conclusion = f"<span style='color:#2ecc71'>✅ Potentially Undervalued:</span> Price ({current_fmt}) suggests **{diff_pct:+.1%}** upside to DCF ({intrinsic_fmt})."
        else: conclusion = f"<span style='color:red'>❌ Valuation Error: Unexpected state.</span>"
    else: conclusion = f"<span style='color:red'>❌ Valuation Error: Final valuation data structure invalid.</span>"
    display(Markdown(f"> **{conclusion}**"))

# --- Execute Summary Dashboard Generation ---
valuation_successful = ( isinstance(final_valuation, dict) and not (final_valuation.get('error') and "Cannot proceed" in final_valuation.get('error','')) and not (final_valuation.get('error') and "prevents valuation" in final_valuation.get('error','')) )

if valuation_successful and isinstance(fcff_results, dict) and isinstance(stock_data.get('info'), dict) and isinstance(input_sources, dict):
    try:
        display_summary_dashboard(
            company_name, ticker_symbol, current_price, final_valuation.get('intrinsic_value_per_share'),
            discount_rate, fcff_results.get('terminal_growth_rate'), fcff_results.get('initial_growth_rate'),
            fcff_results.get('growth_source', 'N/A'),
            final_valuation, stock_data['history'], fcff_results,
            currency_symbol, stock_data['info'], input_sources, is_indian_stock )
    except Exception as e:
         display(Markdown(f"❌ **CRITICAL ERROR during Dashboard Generation: {e}**")); print(traceback.format_exc())
else:
    skip_reason = final_valuation.get('error', 'Unknown upstream error') if isinstance(final_valuation, dict) else "Final valuation data invalid"
    if isinstance(fcff_results,dict) and fcff_results.get('error') and "Cannot proceed" in fcff_results.get('error'): skip_reason = fcff_results.get('error')
    display(Markdown(f"❌ **Skipping Summary Dashboard due to prior errors:** {skip_reason}"));


# %% [markdown]
# ## 11. Assumptions & Disclaimer (Reflects V21.3 Logic)

# %%
# Dynamically generate assumption strings based on final values
rf_source_disp = input_sources.get('Rf', '?')
rm_source_disp = input_sources.get('Rm','?')
beta_source_disp = input_sources.get('Beta','?')
c_spread_source_disp = input_sources.get('Credit Spread','N/A')
bench_source_disp = input_sources.get('Benchmarks','?')
fcff_method_source_disp = input_sources.get('FCFF Method','N/A') # Get selected method source

growth_source_disp = fcff_results.get('growth_source', 'N/A') if isinstance(fcff_results, dict) else 'Error/Not Run'
tax_rate_source_disp = wacc_components.get('tax_rate_source', 'N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
equity_source_disp = wacc_components.get('equity_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
weights_source_disp = wacc_components.get('weights_source','N/A') if isinstance(wacc_components, dict) else 'Error/Not Run'
terminal_growth_final = fcff_results.get('terminal_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan
initial_growth_final = fcff_results.get('initial_growth_rate', np.nan) if isinstance(fcff_results, dict) else np.nan
# FCFF Basis method is now directly from user choice
fcff_basis_method_disp = fcff_results.get('fcff_basis_method', 'N/A') if isinstance(fcff_results, dict) else 'Error/Not Run'


# %%
# ----> UPDATED TEXT TO REFLECT NEW FCFF BASIS LOGIC (V21.3) <----
display(Markdown(f"""
**Key Assumptions & Input Sources:**
*   **Financial Data:** Relies on data provided by Yahoo Finance (`yfinance`). Accuracy subject to provider. Missing data uses fallbacks or causes errors. Assumes annual statements unless only quarterly are available.
*   **Risk-Free Rate (Rf):** Attempted dynamic fetch, **confirmed/overridden by user.** (`Source: {rf_source_disp}`)
*   **Expected Market Return (Rm):** Based on **user input.** (`Source: {rm_source_disp}`)
*   **Market Risk Premium (MRP):** Derived from user `Rm` and final `Rf`.
*   **Beta (β):** Dynamically fetched if valid, otherwise **required user input** (defaults to 1.0 if fetch/input fails). (`Source: {beta_source_disp}`)
*   **Cost of Debt (Kd):** Implied from Interest Exp/Total Debt if possible & within bounds (`{format_percent(MIN_COST_OF_DEBT)}` - `{format_percent(MAX_COST_OF_DEBT)}`). Otherwise, defaults to `Rf + Credit Spread` (bounded).
*   **Credit Spread:** Used only if default Kd triggered. Based on **user input.** (`Source: {c_spread_source_disp}`)
*   **FCFF Calculation Method:** **Selected by user upfront.** (`Source: {fcff_method_source_disp}`).
    *   Historical FCFF (T-0 to T-3) is calculated using **only the selected method's formula.**
    *   The **4-year average** of these historical FCFFs is used as the projection starting basis.
    *   **Projection halts if this 4-year average is not positive.** No fallback methods are attempted for the basis.
*   **Growth Rates:**
    *   **Initial Growth:** `{format_percent(initial_growth_final)}` used for Year 1 (`Source: {growth_source_disp}`)
        1.  Fetch last 4 data points (T-0 to T-3) of Revenue, FCFF (**using the user-selected method**), and Stock Price.
        2.  Calculate **3-Year CAGR** (T-0 vs T-3) for Rev, Stock, and FCFF (selected method) where start data is valid & positive.
        3.  Calculate the **Geometric Mean** of successfully calculated, non-extreme CAGRs.
        4.  If GM fails or no valid CAGRs, use **Default (`{format_percent(DEFAULT_GROWTH_RATE)}`)**.
        5.  Apply **Bounds:** Cap between **`{format_percent(MIN_INITIAL_GROWTH_FLOOR)}`** and **`{format_percent(MAX_INITIAL_GROWTH_CEILING)}`**.
    *   **Terminal Growth (g):** `{format_percent(terminal_growth_final)}` = min(GDP cap, WACC*`0.60`, Final Initial Growth, Rf*`0.90`), floored at `{format_percent(MIN_GROWTH)}`, ensured `< WACC`.
*   **Projection Period:** FCFF is projected for **{FORECAST_YEARS} years.**
*   **Tax Rate:** Effective rate derived from latest financials ({tax_rate_source_disp}) (bounded) or default (`{format_percent(DEFAULT_TAX_RATE)}`).
*   **Capital Structure:** Assumed stable using `{equity_source_disp}` and calculated debt. (`Weights Source: {weights_source_disp}`)
*   **Industry Benchmarks:** Uses **user input** or **generic fallbacks.** (`Source: {bench_source_disp}`)
*   **Currency Formatting:** Standard K/M/B/T or ₹/Lakh/Cr based on ticker.

**Limitations:**
*   **Model Sensitivity:** DCF highly sensitive to inputs (WACC, growth rates, **chosen FCFF method**).
*   **FCFF Method Choice Impact:** The choice of FCFF calculation method can significantly alter historical values, the average basis, FCFF CAGR, and thus the final valuation. Method 4 (EBITDA-Tax) is a simplification.
*   **No Fallbacks:** If the chosen FCFF method results in missing components or a non-positive 4yr average basis, the valuation will halt.
*   **Input Quality:** Validity depends on user inputs and appropriateness of fallbacks. Benchmarks should be carefully chosen.
*   **Historical Data Availability:** Limited statement data impacts CAGR and average basis reliability.
*   **CAGR/GM Modifications:** Standard limitations apply (positive start needed for CAGR, GM sensitive to large negatives).
*   **Dynamic Data Reliability:** Depends on `yfinance` accuracy/availability.
*   **Simplifications:** Linear growth decay, constant tax rate/structure in projection. Ignores qualitative factors, macro shocks, etc.
*   **Fallback Accuracy:** Default growth rate/benchmarks may not be representative.
*   **Data Completeness:** Performance degrades with missing data. *Cross-verify key data.*

**Disclaimer:**
*For **educational and illustrative purposes only**. **NOT investment advice**. Models have limitations. Users MUST conduct own due diligence and consult a qualified financial advisor before investing. Creators NOT liable for actions based on this report.*
"""))


# %%






````

````python
# %% [markdown]
# # Dynamic Stock Financial Analysis & DCF Valuation
# **(Interactive Report - Dynamic/User Input Driven Assumptions - V21.6 - Syntax Fix / Consistent Indian Formatting)**
#
# **Analyst:** Simulated AI Research Analyst
#
# **Objective:** To perform a comprehensive financial analysis and estimate the intrinsic value per share using DCF. User selects FCFF method. **Uses consistent Lakh/Crore formatting.** Initial growth: GM of Rev/FCFF CAGR. **Syntax error in WACC fixed.** **Syntax error in safe_get fixed.**
#
# **Methodology:**
# 1.  User Input & Dynamic Fetching.
# 2.  Data Acquisition.
# 3.  WACC Calculation (**Syntax Fix Applied**).
# 4.  FCFF Method Pre-Calculation & User Selection (Displays Avg FCFF).
# 5.  FCFF Calculation & Projection (3 Years, Selected Method Only, Halts if basis non-positive).
# 6.  Terminal Value.
# 7.  Discounting & Valuation.
# 8.  Ratio Analysis (Plots fixed).
# 9.  Interactive Visualization.
# 10. **Consistent Indian Number Formatting (Lakh/Crore) enforced.**

# %% [markdown]
# ## 1. Setup and Libraries
# (Section Unchanged)

# %%
# Standard Libraries
import warnings
import sys
import numpy as np
import pandas as pd
import datetime
import math # For isnan, isinf checks
import traceback # For detailed error logging

# Data Fetching
import yfinance as yf

# Visualization (Using Plotly)
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio

# Display Utilities
from IPython.display import display, Markdown, HTML

# Settings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pio.templates.default = "plotly_white"

# %% [markdown]
# ## 2. Configuration and Helper Functions (Formatting Updated V21.6)

# %%
# --- Model Configuration ---
DEFAULT_TICKER = "MUTHOOTFIN.NS"
FORECAST_YEARS = 3
TERMINAL_GROWTH_RATE_CAP = 0.05
DEFAULT_GROWTH_RATE = 0.065
DEFAULT_RF_RATE_SUGGESTION = 7.0
DEFAULT_CURRENCY_SYMBOL = "₹" # Set to Rupee by default
DEFAULT_TAX_RATE = 0.25
DEFAULT_CREDIT_SPREAD = 0.015
DEFAULT_MARKET_RISK_PREMIUM = 0.06

# --- Generic Fallback Benchmark Ratios ---
DEFAULT_BENCHMARK_RATIOS = {
    'P/E': 17.3, 'P/B': 2.5, 'EV/EBITDA': 11.0, 'ROE': 19.05,
    'ROA': 3.92, 'ROCE': 11.85,'Asset Turnover': 0.15,
    'Current Ratio': 1.5, 'Quick Ratio': 1.5
}

# --- Model Stability Bounds ---
MIN_WACC = 0.06; MAX_WACC = 0.25; MIN_TAX_RATE = 0.10; MAX_TAX_RATE = 0.40
MIN_GROWTH = 0.00; MIN_INITIAL_GROWTH_FLOOR = 0.011; MAX_INITIAL_GROWTH_CEILING = 0.15
MIN_COST_OF_DEBT = 0.03; MAX_COST_OF_DEBT = 0.18

# --- Ratio Keys Needed for Benchmarking ---
REQUIRED_BENCHMARK_RATIOS = list(DEFAULT_BENCHMARK_RATIOS.keys())

# --- FCFF Method Definitions ---
FCFF_METHODS_DEFINITION = {
    0: {"name": "EBITDA - TaxProv - WC - Capex", "formula": "EBITDA - Tax Provision - Change in WC - CapEx"},
    1: {"name": "NOPAT + Depr - WC - Capex", "formula": "EBIT*(1-Tax Rate) + Depr&Amort - Change in WC - CapEx"},
    2: {"name": "CFO + IntExp*(1-T) - Capex", "formula": "Cash Flow Ops + Interest Exp*(1-Tax Rate) - CapEx"},
    3: {"name": "CFO - Capex", "formula": "Cash Flow Ops - CapEx"},
    4: {"name": "EBITDA - TaxProv (Proxy)", "formula": "EBITDA - Tax Provision (Simplified Proxy)"}
}


# --- Helper Functions ---
def print_keys(statement, name): # (Unchanged)
    print(f"\n--- Keys in {name} ---");
    if isinstance(statement, pd.DataFrame): print(", ".join(map(str, [idx for idx in statement.index if idx is not None])))
    elif isinstance(statement, dict): print(", ".join(statement.keys()))
    print("--------------------\n")

# *** SYNTAX FIX APPLIED HERE V21.6 ***
# Corrected try/except structure and conversion logic
def safe_get(data, key_list, statement_name="Statement", default=np.nan):
    """Safely retrieves and attempts to convert financial data to float."""
    if not isinstance(key_list, list): key_list = [key_list]
    value = None
    found_key = None

    for key in key_list:
        current_value = None
        raw_val = None
        data_source = None # Keep track of where we might find it

        try: # Outer try: Getting the raw value
            if isinstance(data, pd.Series):
                data_source = f"'{statement_name}' Series"
                if key in data.index:
                    raw_val = data.loc[key]
                # If key not in index, raw_val remains None, handled later
            elif isinstance(data, dict):
                data_source = f"'{statement_name}' Dict"
                raw_val = data.get(key) # dict.get handles missing keys gracefully (returns None)
            else:
                # Handle unexpected data types if necessary, or just let raw_val be None
                pass # Raw_val remains None

            # Inner processing: Converting raw_val if found
            if raw_val is not None:
                # Check for explicit non-values first
                if isinstance(raw_val, str) and raw_val.lower() in ['n/a', 'none', '', '--', '-']:
                     pass # Treat as missing, current_value remains None
                # Handle direct numeric types
                elif isinstance(raw_val, (int, float)):
                    if np.isfinite(raw_val):
                        current_value = float(raw_val)
                # Handle single-value Pandas Series
                elif isinstance(raw_val, pd.Series):
                     if not raw_val.empty:
                         series_val = raw_val.iloc[0]
                         # Check if the value extracted from Series is numeric and finite
                         if isinstance(series_val, (int, float)) and np.isfinite(series_val):
                             current_value = float(series_val)
                         elif pd.notna(series_val): # Check for NaN explicitly in Series context
                             try:
                                 temp_val = float(series_val)
                                 if np.isfinite(temp_val):
                                     current_value = temp_val
                             except (ValueError, TypeError):
                                 pass # Conversion failed
                 # Try converting other types (like strings representing numbers)
                else:
                    try: # Inner-inner try: Convert potentially non-numeric to float
                        temp_val = float(raw_val)
                        if np.isfinite(temp_val):
                            current_value = temp_val
                    except (ValueError, TypeError):
                        pass # Conversion failed, current_value remains None

        except Exception as e: # Catch errors during data access (e.g., unexpected Series issues, not KeyError which is handled by 'in data.index')
            # print(f"Warning in safe_get access for key '{key}' in {statement_name}: {e}") # Optional detailed logging
            pass # Let current_value remain None

        # Check if we found a valid (finite numeric) value in this iteration
        if current_value is not None:
            value = current_value
            found_key = key
            # print(f"Debug: Found key '{found_key}' from {data_source}, value: {value}") # Optional debug print
            break # Exit the loop once a valid value is found

    # Final fallback to default if no valid value was found across all keys
    if value is None:
        value = default
        # print(f"Debug: Key(s) {key_list} not found or invalid in {statement_name}, using default: {value}") # Optional debug print

    return value

# MODIFIED format_currency to ALWAYS use Indian Lakh/Crore format
def format_currency(value, symbol="₹", precision=2):
    """Formats currency value using Indian Lakh/Crore notation ONLY."""
    # Overriding symbol internally to always be Rupee
    symbol = "₹"
    if value is None or pd.isna(value) or not math.isfinite(value): return "N/A"
    if abs(value) > 1e18: return "Inf" # Handle potential overflows
    sign = "-" if value < 0 else ""; abs_val = abs(value)

    if abs_val >= 1e7: # Crores
        formatted_val = f"{abs_val / 1e7:.{precision}f} Cr"
    elif abs_val >= 1e5: # Lakhs
        formatted_val = f"{abs_val / 1e5:.{precision}f} Lakh"
    else: # Base value
        # Ensure comma separation for values below Lakh
        formatted_val = f"{abs_val:,.{precision}f}"

    return f"{sign}{symbol} {formatted_val}"


def format_percent(value, precision=1): # (Unchanged)
     if value is None or pd.isna(value) or not math.isfinite(value): return "N/A"
     if abs(value) > 1e5: return "Inf"
     return f"{value * 100:.{precision}f}%"

def get_validated_float_input(prompt_message, default_value=None, allow_blank=False, min_val=None, max_val=None, is_percentage=False): # (Unchanged)
    while True:
        prompt_with_default = prompt_message
        if is_percentage: prompt_with_default += " (e.g., '15' for 15%)"
        if default_value is not None: prompt_with_default += f" [Default: {default_value}]: "
        elif allow_blank: prompt_with_default += f" [Leave blank]: "
        else: prompt_with_default += ": "
        try:
            user_input_str = input(prompt_with_default).strip()
            if not user_input_str:
                if allow_blank: return None
                if default_value is not None: return float(default_value)
                raise ValueError("Input cannot be empty.")
            if user_input_str.endswith('%'): user_input_str = user_input_str[:-1]
            value = float(user_input_str)
            if min_val is not None and value < min_val: raise ValueError(f"Value must be >= {min_val}")
            if max_val is not None and value > max_val: raise ValueError(f"Value must be <= {max_val}")
            # If percentage, the input is number like 15, not 0.15 yet. Return as number.
            return value
        except ValueError as e: print(f"❌ Invalid input: {e}. Please enter a valid number within the allowed range.")


def try_get_dynamic_rf(rf_ticker_symbol="^TNX"): # (Unchanged)
    print(f"Attempting fetch: Rf ({rf_ticker_symbol})...")
    try:
        rf_ticker=yf.Ticker(rf_ticker_symbol); hist=rf_ticker.history(period="5d")
        if not hist.empty and 'Close' in hist.columns:
            last_close=hist['Close'].iloc[-1]
            if pd.notna(last_close):
                dynamic_rf=last_close/100.0 # Assume ticker value is percentage points
                if 0.001 < dynamic_rf < 0.20: # Reasonable bounds for Rf rate
                    print(f"✅ Dynamic Rf fetched: {format_percent(dynamic_rf)}")
                    return dynamic_rf
                else:
                    print(f"⚠️ Dynamic Rf ({format_percent(dynamic_rf)}) out of expected range [0.1%, 20%]. Ignoring.")
                    return None
            else: print(f"⚠️ Rf data contains NaN."); return None
        else: print(f"⚠️ No recent history data for Rf ticker."); return None
    except Exception as e: print(f"❌ Error fetching Rf rate for {rf_ticker_symbol}: {e}"); return None

# --- Growth Calculation Helpers (Unchanged) ---
def calculate_cagr_modified(start_val, end_val, num_years): # (Unchanged)
    if pd.isna(start_val) or pd.isna(end_val) or not math.isfinite(start_val) or not math.isfinite(end_val) or num_years <= 0: return np.nan
    # Handle non-positive start value gracefully
    if start_val <= 1e-9:
        # print("CAGR Warning: Start value is non-positive.")
        return np.nan # Cannot calculate meaningful CAGR

    cagr = np.nan
    try:
        # Handle non-positive end value - cannot calculate standard CAGR directly
        if end_val <= 1e-9:
            # print("CAGR Warning: End value is non-positive.")
            # Could return a large negative number or NaN depending on interpretation
            return np.nan # Sticking to NaN for undefined growth
        else:
            ratio = end_val / start_val
            # Ensure ratio is positive before taking root
            if ratio > 0:
                cagr = ratio**(1 / num_years) - 1
            else:
                # This case shouldn't happen if start_val > 0 and end_val > 0
                 return np.nan

    except Exception as e:
        # print(f"CAGR Calculation Error: {e}")
        cagr = np.nan

    # Add bounds check for plausibility
    if pd.notna(cagr) and not (-0.999 < cagr < 20.0): # Limit growth rates from -99.9% to +2000%
        # print(f"CAGR Warning: Result {format_percent(cagr)} out of plausible bounds [-99.9%, +2000%]. Setting to NaN.")
        cagr = np.nan

    return cagr

def geometric_mean_modified(rates): # (Unchanged)
    valid_rates = [r for r in rates if pd.notna(r) and math.isfinite(r)]
    if not valid_rates: return np.nan
    if len(valid_rates) == 1: return valid_rates[0]

    offset = 1.0
    factors_offset = [r + offset for r in valid_rates]

    # Check if any rate implies a negative factor (rate <= -100%)
    if any(f <= 1e-9 for f in factors_offset):
        print("⚠️ GM calculation failure (rate <= -100%). Using arithmetic mean as fallback.")
        return np.mean(valid_rates)

    n = len(valid_rates)
    product = 1.0
    try:
        for factor in factors_offset:
            product *= factor

        # Handle potential complex number if product is negative and n is even
        if product < 0 and n % 2 == 0:
            print("⚠️ GM calculation failure (negative base with even root). Using arithmetic mean as fallback.")
            return np.mean(valid_rates)

        # Calculate geometric mean with offset
        gm_offset = np.sign(product) * (np.abs(product)**(1 / n))
        gm = gm_offset - offset

        # Check plausibility bounds for the result
        if pd.notna(gm) and not (-0.999 < gm < 20.0):
            print(f"⚠️ GM result ({format_percent(gm)}) out of bounds [-99.9%, +2000%]. Using arithmetic mean.")
            return np.mean(valid_rates)

        return gm

    except Exception as e:
        print(f"⚠️ GM numeric error ({e}). Using arithmetic mean as fallback.")
        return np.mean(valid_rates)


# --- FCFF Component Fetcher (Unchanged - Uses updated safe_get and format_currency) ---
def _get_fcff_components(fin_col, bs_col, cf_col, bs_prev_col, log_list):
    components = {}; log_list.append("--- Fetching FCFF Components ---")
    rev_keys=["Total Revenue","Revenue"]; components['revenue']=safe_get(fin_col, rev_keys, "IS", default=np.nan); log_list.append(f"    Revenue: {format_currency(components['revenue'])}") # format_currency handles formatting
    ebit_keys=["EBIT","Ebit","Operating Income"]; components['ebit']=safe_get(fin_col, ebit_keys, "IS", default=np.nan)
    # Try deriving EBIT if direct fetch fails
    if not np.isfinite(components['ebit']):
        pretax=safe_get(fin_col, ['Pretax Income'], "IS(Fallback)", default=np.nan); int_exp=abs(safe_get(fin_col, ['Interest Expense', 'Interest Expense Non Operating'], "IS(Fallback)", default=np.nan))
        if np.isfinite(pretax) and np.isfinite(int_exp):
            components['ebit']=pretax+int_exp; log_list.append(f"    EBIT (Derived from Pretax+Interest): {format_currency(components['ebit'])}")
        else:
            log_list.append(f"    EBIT: N/A (Direct and Derived failed)")
            components['ebit']=np.nan
    else: log_list.append(f"    EBIT: {format_currency(components['ebit'])}")

    dep_keys=["Depreciation And Amortization","Depreciation"]; components['dep_amort']=safe_get(cf_col, dep_keys, "CF", default=np.nan)
    # Fallback to Income Statement if not found in Cash Flow
    if not np.isfinite(components['dep_amort']):
        components['dep_amort']=safe_get(fin_col, dep_keys, "IS(Fallback)", default=np.nan)
        if np.isfinite(components['dep_amort']): log_list.append(f"    Dep&Amort (from IS): {format_currency(components['dep_amort'])}")
        else: log_list.append(f"    Dep&Amort: N/A (CF and IS failed)")
    else: log_list.append(f"    Dep&Amort (from CF): {format_currency(components['dep_amort'])}")

    # Calculate EBITDA = EBIT + Dep&Amort
    if np.isfinite(components.get('ebit')) and np.isfinite(components.get('dep_amort')):
        components['ebitda']=components['ebit']+components['dep_amort']; log_list.append(f"    EBITDA (Calculated): {format_currency(components['ebitda'])}")
    else: components['ebitda']=np.nan; log_list.append(f"    EBITDA: N/A (Missing EBIT or Dep&Amort)")

    capex_keys=["Capital Expenditure","Net Capital Expenditure", "Purchase Of Property Plant Equipment"]; capex_val=safe_get(cf_col, capex_keys, "CF", default=np.nan); components['capex']=abs(capex_val) if np.isfinite(capex_val) else np.nan; log_list.append(f"    CapEx (Abs): {format_currency(components['capex'])}")

    wc_keys=["Change In Working Capital", "Changes In Working Capital"]; components['wc_cf']=safe_get(cf_col, wc_keys, "CF(WC)", default=np.nan)
    # Use direct WC change from CF if available
    if np.isfinite(components['wc_cf']):
        components['wc']=components['wc_cf']; components['wc_src']="CF Direct"; log_list.append(f"    WC Change (from CF Direct): {format_currency(components['wc'])}")
    else:
        # Try calculating from Balance Sheet if direct CF value missing or previous BS exists
        components['wc_src']="BS Calc Attempt"; components['wc']=np.nan
        log_list.append("    WC Change (CF Direct not found, trying BS calc...)")
        # Need both current and previous balance sheets
        if isinstance(bs_prev_col, pd.Series) and isinstance(bs_col, pd.Series):
            curr_a=safe_get(bs_col, ['Current Assets', 'Total Current Assets'], 'BS Current',np.nan)
            curr_l=safe_get(bs_col, ['Current Liabilities', 'Total Current Liabilities'], 'BS Current',np.nan)
            prev_a=safe_get(bs_prev_col, ['Current Assets', 'Total Current Assets'], 'BS Previous',np.nan)
            prev_l=safe_get(bs_prev_col, ['Current Liabilities', 'Total Current Liabilities'], 'BS Previous',np.nan)

            if all(map(np.isfinite, [curr_a, curr_l, prev_a, prev_l])):
                current_nwc = curr_a - curr_l
                previous_nwc = prev_a - prev_l
                components['wc'] = current_nwc - previous_nwc # Change = Current WC - Previous WC
                components['wc_src']="BS Calc Success"
                log_list.append(f"      Current NWC: {format_currency(current_nwc)} | Previous NWC: {format_currency(previous_nwc)}")
                log_list.append(f"    WC Change (BS Calculated): {format_currency(components['wc'])}")
            else:
                components['wc_src']="BS Calc Fail (Missing Components)"
                log_list.append("    WC Change (BS Calc): Failed - Missing BS components")
                # Log which components might be missing
                missing_bs = []
                if not np.isfinite(curr_a): missing_bs.append("Current Assets")
                if not np.isfinite(curr_l): missing_bs.append("Current Liabilities")
                if not np.isfinite(prev_a): missing_bs.append("Previous Assets")
                if not np.isfinite(prev_l): missing_bs.append("Previous Liabilities")
                if missing_bs: log_list.append(f"      Missing: {', '.join(missing_bs)}")
        else:
            components['wc_src']="BS Calc Fail (No Prev BS)"
            log_list.append("    WC Change (BS Calc): Failed - Previous Balance Sheet data missing")

    cfo_keys=["Cash Flow From Operating Activities", "Operating Cash Flow"]; components['cfo']=safe_get(cf_col, cfo_keys, "CF", default=np.nan); log_list.append(f"    CFO: {format_currency(components['cfo'])}")
    int_exp_keys=['Interest Expense', 'Interest Expense Non Operating']; components['int_exp']=abs(safe_get(fin_col, int_exp_keys, "IS", default=np.nan)); log_list.append(f"    Interest Exp (Abs): {format_currency(components['int_exp'])}")
    tax_keys=['Tax Provision', 'Income Tax Paid', 'Income Tax Expense']; components['tax_prov']=safe_get(fin_col, tax_keys, "IS", default=np.nan)
    # Fallback for Tax if not on IS
    if not np.isfinite(components['tax_prov']):
        components['tax_prov']=safe_get(cf_col, tax_keys, "CF(Fallback)", default=np.nan)
        if np.isfinite(components['tax_prov']): log_list.append(f"    Tax Prov (from CF): {format_currency(components['tax_prov'])}")
        else: log_list.append(f"    Tax Prov: N/A (IS and CF failed)")
    else: log_list.append(f"    Tax Prov (from IS): {format_currency(components['tax_prov'])}")

    return components

# --- FCFF Calculation Helper: Single Method (Unchanged - Uses updated _get_fcff_components) ---
def calculate_fcff_single_method_single_year(method_choice, fin_col, bs_col, cf_col, bs_prev_col, tax_rt, log_list_parent):
    log_list_local=[]; c=_get_fcff_components(fin_col, bs_col, cf_col, bs_prev_col, log_list_local); log_list_parent.extend(log_list_local)
    fcff_val=np.nan; method_name=FCFF_METHODS_DEFINITION.get(method_choice,{}).get("name","?"); log_list_parent.append(f"--- Calc FCFF: {method_name} ---")
    valid_tax=np.isfinite(tax_rt);
    if not valid_tax and method_choice in [1,2]: log_list_parent.append(f"    ❌ Invalid Tax Rate ({tax_rt}) needed for method {method_choice}. Cannot calculate."); return np.nan

    # Check for required components based on method
    required_comps = {}
    if method_choice == 0: required_comps = {'ebitda': c.get('ebitda'), 'tax_prov': c.get('tax_prov'), 'wc': c.get('wc'), 'capex': c.get('capex')}
    elif method_choice == 1: required_comps = {'ebit': c.get('ebit'), 'dep_amort': c.get('dep_amort'), 'wc': c.get('wc'), 'capex': c.get('capex')}
    elif method_choice == 2: required_comps = {'cfo': c.get('cfo'), 'int_exp': c.get('int_exp'), 'capex': c.get('capex')}
    elif method_choice == 3: required_comps = {'cfo': c.get('cfo'), 'capex': c.get('capex')}
    elif method_choice == 4: required_comps = {'ebitda': c.get('ebitda'), 'tax_prov': c.get('tax_prov')}
    else: log_list_parent.append(f"    ❌ Invalid method choice: {method_choice}."); return np.nan

    missing_comps_list = [name for name, val in required_comps.items() if not np.isfinite(val)]
    if missing_comps_list:
        log_list_parent.append(f"    ❌ Missing Component(s) for Method {method_choice}: {', '.join(missing_comps_list)}")
        return np.nan

    # Perform calculation if all components are valid
    try:
        if method_choice==0: #EBITDA-Tax-WC-Capex
            fcff_val=c['ebitda']-c['tax_prov']-c['wc']-c['capex']
        elif method_choice==1: #NOPAT+D-WC-C
            fcff_val=c['ebit']*(1-tax_rt)+c['dep_amort']-c['capex']-c['wc']
        elif method_choice==2: #CFO+Int(1-T)-C
            fcff_val=c['cfo']-c['capex']+c['int_exp']*(1-tax_rt)
        elif method_choice==3: #CFO-C
            fcff_val=c['cfo']-c['capex']
        elif method_choice==4: #EBITDA-Tax (Proxy)
            fcff_val=c['ebitda']-c['tax_prov']

        if np.isfinite(fcff_val): log_list_parent.append(f"    Calc FCFF = {format_currency(fcff_val)}") # Use updated formatter
        else: log_list_parent.append(f"    Calculation resulted in Non-Finite value.")

    except Exception as e: log_list_parent.append(f"    ❌ Calculation Error: {e}"); print(traceback.format_exc()); return np.nan

    # Return NaN if calculation wasn't successful or resulted in NaN/inf
    return fcff_val if np.isfinite(fcff_val) else np.nan


# --- End Helper Functions ---

# %% [markdown]
# ## 3. User Input & Dynamic Fetching: Ticker, Rf, Rm, Benchmarks, Beta, Spread
# (Section Unchanged - Uses updated get_validated_float_input and format_percent)

# %%
input_sources = {}
user_ticker = input(f"Enter ticker [Default: {DEFAULT_TICKER}]: ") or DEFAULT_TICKER
ticker_symbol = user_ticker.strip().upper(); display(Markdown(f"**Ticker:** `{ticker_symbol}`"))

is_indian_stock = ticker_symbol.endswith((".NS", ".BO"))
if is_indian_stock: print("Info: Indian stock detected (.NS or .BO suffix).")
# Suggest appropriate Rf proxy based on market
rf_proxy_ticker = "^NS10" if is_indian_stock else "^TNX"; print(f"Info: Using suggested Rf proxy for dynamic fetch: {rf_proxy_ticker}")

print("\n--- CAPM Inputs ---")
dynamic_rf_rate = try_get_dynamic_rf(rf_proxy_ticker) # Returns rate (e.g., 0.07) or None

# Use dynamic Rf if available and valid, otherwise use default suggestion
suggested_rf_val = dynamic_rf_rate * 100 if dynamic_rf_rate else DEFAULT_RF_RATE_SUGGESTION
rf_prompt = f"Enter Risk-Free Rate (Rf) %"
# Get input as percentage value (e.g., user enters '7' for 7%)
risk_free_rate_input_pct = get_validated_float_input(rf_prompt, default_value=f"{suggested_rf_val:.2f}", min_val=0.0, max_val=20.0, is_percentage=True)
risk_free_rate_input = risk_free_rate_input_pct / 100.0 # Convert to rate (e.g., 0.07)
input_sources['Rf'] = 'Dynamic Fetch' if dynamic_rf_rate and abs(risk_free_rate_input - dynamic_rf_rate) < 1e-4 else 'User Input/Default'
display(Markdown(f"**Rf:** {format_percent(risk_free_rate_input)} (`Source: {input_sources['Rf']}`)"))

# Market Return (Rm) - suggest based on Rf + default MRP
suggested_mrp = DEFAULT_MARKET_RISK_PREMIUM
suggested_rm = risk_free_rate_input + suggested_mrp
rm_prompt = f"Enter Expected Market Return (Rm) %. Suggestion based on Rf + {format_percent(suggested_mrp)} MRP: {format_percent(suggested_rm)}"
while True:
    market_return_input_pct = get_validated_float_input(rm_prompt, default_value=f"{suggested_rm*100:.1f}", min_val=-10.0, max_val=35.0, is_percentage=True)
    market_return_input = market_return_input_pct / 100.0 # Convert to rate
    if market_return_input >= risk_free_rate_input:
        break
    else: print(f"❌ Market Return (Rm) {format_percent(market_return_input)} must be greater than or equal to Risk-Free Rate (Rf) {format_percent(risk_free_rate_input)}. Please re-enter.")

market_risk_premium_input = market_return_input - risk_free_rate_input
input_sources['Rm'] = 'User Input/Default'
input_sources['MRP'] = 'Derived (Rm - Rf)'
display(Markdown(f"**Rm:** {format_percent(market_return_input)} (`Source: {input_sources['Rm']}`)"))
display(Markdown(f"**MRP:** {format_percent(market_risk_premium_input)} (`Source: {input_sources['MRP']}`)"))

# Initialize Beta and Credit Spread (will be fetched/asked for later if needed)
beta_input = None; input_sources['Beta'] = 'Pending Fetch/Input'
user_credit_spread = None; input_sources['Credit Spread'] = 'Pending WACC Calc'

print("\n--- Ratio Benchmarks ---"); display(Markdown("Enter benchmark values for comparison, or leave blank to use defaults."))
user_benchmarks={}; benchmark_sources={}; print("Provide benchmark ratios:")
for ratio_key in REQUIRED_BENCHMARK_RATIOS:
    default_val = DEFAULT_BENCHMARK_RATIOS.get(ratio_key)
    prompt = f"  Benchmark {ratio_key}"
    # Determine if the ratio is typically expressed as a percentage
    is_perc = ratio_key in ['ROE','ROA','ROCE']
    # Set reasonable min/max bounds for input validation
    min_b, max_b = (-100, 200) if is_perc else (-100, 500) # Wider range for non-percentage ratios

    # Get validated input (allow blank to use default)
    # Note: get_validated_float_input returns the number as entered (e.g., 15 for 15%)
    bench_val_input = get_validated_float_input(prompt, default_value=f"{default_val:.2f}" if default_val is not None else None, allow_blank=True, min_val=min_b, max_val=max_b, is_percentage=is_perc)

    final_bench_val = np.nan # Initialize as NaN
    source = 'N/A'

    if bench_val_input is not None: # User provided a value
        # Convert to rate if it was a percentage input
        final_bench_val = bench_val_input / 100.0 if is_perc else bench_val_input
        user_benchmarks[ratio_key] = final_bench_val
        benchmark_sources[ratio_key] = 'User Input'
        source = 'User Input'
    elif default_val is not None: # User left blank, use default if available
        # Defaults are already stored as rates (e.g., ROE=19.05 means 0.1905)
        final_bench_val = default_val / 100.0 if is_perc else default_val # Ensure default is also treated as rate if perc
        user_benchmarks[ratio_key] = final_bench_val
        benchmark_sources[ratio_key] = 'Default'
        source = 'Default'
    else: # User left blank, no default available
        user_benchmarks[ratio_key] = np.nan
        benchmark_sources[ratio_key] = 'N/A'
        source = 'N/A'

    # Display the final benchmark value being used
    disp_val = "N/A"
    if np.isfinite(final_bench_val):
        # Use format_percent for ROE/ROA/ROCE, otherwise format as number
        disp_val = format_percent(final_bench_val) if is_perc else f"{final_bench_val:.2f}"

    display(Markdown(f"  - **{ratio_key}:** {disp_val} (`Source: {source}`)"))

input_sources['Benchmarks'] = 'User/Default Mix'


# %% [markdown]
# ## 4. Data Acquisition (Fetching Extended History)
# (Section Unchanged - Uses updated safe_get and format_currency)

# %%
stock_data = {}
try:
    stock = yf.Ticker(ticker_symbol)
    # Attempt to fetch basic info first
    info = stock.info
    # Check if info is valid and contains a price; if not, try history as a fallback check
    if not info or not info.get('regularMarketPrice'):
        print(f"Warning: Basic info dictionary seems limited or missing price for {ticker_symbol}. Attempting to fetch history to confirm ticker validity...")
        hist_check = stock.history(period="1mo") # Fetch a short period just to see if ticker works
        if hist_check.empty:
            raise ValueError(f"Ticker '{ticker_symbol}' might be invalid or delisted. No basic info or recent history found.")
        else:
             print(f"Info: History data found for {ticker_symbol}. Proceeding, but some info fields might be missing.")
             info = {} if not isinstance(info, dict) else info # Ensure info is a dict, even if empty

    stock_data['info'] = info

    # Fetch financial statements (annual first, fallback to quarterly)
    stock_data['financials'] = stock.financials
    stock_data['balance_sheet'] = stock.balance_sheet
    stock_data['cashflow'] = stock.cashflow

    if stock_data['financials'].empty: print("Warning: Annual Income Statement empty, trying quarterly financials..."); stock_data['financials']=stock.quarterly_financials
    if stock_data['balance_sheet'].empty: print("Warning: Annual Balance Sheet empty, trying quarterly balance sheet..."); stock_data['balance_sheet']=stock.quarterly_balance_sheet
    if stock_data['cashflow'].empty: print("Warning: Annual Cash Flow statement empty, trying quarterly cashflow..."); stock_data['cashflow']=stock.quarterly_cashflow

    # Check again after fetching quarterly data
    missing_statements = [s for s in ['financials','balance_sheet','cashflow'] if stock_data.get(s, pd.DataFrame()).empty]
    if missing_statements:
        print(f"⚠️ CRITICAL WARNING: Missing essential financial statement(s): {', '.join(missing_statements)}. DCF and Ratio analysis will likely fail or be highly unreliable.")
        # Decide whether to exit or proceed with caution
        # sys.exit(f"Exiting due to missing critical data: {', '.join(missing_statements)}")

    # Fetch historical price data
    stock_data['history'] = stock.history(period="2y") # For recent trends, maybe beta calc if needed
    stock_data['history_long'] = stock.history(period="4y", interval="1d") # For growth calcs / context

    if stock_data['history_long'] is not None and not stock_data['history_long'].empty:
        print(f"Info: Fetched long price history ({len(stock_data['history_long'])} days over approx 4 years).")
        if len(stock_data['history_long']) < 2: print(f"⚠️ Warning: Long history has fewer than 2 data points.")
    else: print(f"⚠️ Warning: Could not fetch long price history (4y).")

    # Extract key info, with fallbacks
    company_name = info.get("longName", info.get("shortName", ticker_symbol)) # Use shortName if longName missing
    currency_code = info.get("currency", DEFAULT_CURRENCY_SYMBOL)
    if not isinstance(currency_code, str) or len(currency_code) > 5: currency_code = DEFAULT_CURRENCY_SYMBOL # Sanitize currency
    currency_symbol = "₹" # Forcing Rupee symbol due to V21.6 formatting requirement

    # Get current price with fallbacks
    price_keys = ['currentPrice', 'regularMarketPrice', 'previousClose', 'regularMarketPreviousClose']
    current_price = safe_get(info, price_keys, 'Info', default=np.nan)
    price_source = "Info (current/regular/previous)" if np.isfinite(current_price) else "N/A from Info"

    # Fallback to history if info price is missing
    if not np.isfinite(current_price) and stock_data['history'] is not None and not stock_data['history'].empty:
        if 'Close' in stock_data['history'].columns and not stock_data['history']['Close'].empty:
            last_close = stock_data['history']['Close'].iloc[-1]
            if pd.notna(last_close) and last_close > 0:
                current_price = last_close
                price_source = "History (Last Close)"
                print(f"Info: Used last closing price from history as current price.")
            else: price_source += ", History Invalid"
        else: price_source += ", History Missing Close"
    else: price_source += ", History Not Checked"


    if not np.isfinite(current_price):
        print(f"⚠️ WARNING: Could not determine current price for {ticker_symbol}.")
        current_price = np.nan # Ensure it's NaN if not found

    # Get Shares Outstanding with fallbacks
    shares_outstanding = safe_get(info, ['sharesOutstanding'], 'Info', default=0)
    shares_source = "Info"
    if not np.isfinite(shares_outstanding) or shares_outstanding <= 0:
        print("Warning: Shares outstanding invalid or missing from info. Trying Balance Sheet...")
        shares_source = "Info Invalid"
        # Fallback to Balance Sheet (look for 'Share Issued' or similar)
        if stock_data['balance_sheet'] is not None and not stock_data['balance_sheet'].empty:
            # Check the most recent column (index 0)
            bs_latest = stock_data['balance_sheet'].iloc[:, 0]
            # Common keys for shares issued/outstanding on BS
            shares_bs_keys = ['Share Issued', 'Ordinary Shares Number', 'Total Common Shares Outstanding']
            shares_bs = safe_get(bs_latest, shares_bs_keys, 'Balance Sheet', 0)
            if shares_bs is not None and np.isfinite(shares_bs) and shares_bs > 0:
                shares_outstanding = shares_bs
                shares_source = "Balance Sheet"
                print(f"Info: Using shares outstanding from Balance Sheet: {shares_outstanding:,.0f}")
            else:
                print("Warning: Could not find valid shares outstanding on the Balance Sheet.")
                shares_source += ", BS Invalid/Missing"
                shares_outstanding = 0 # Keep as 0 or NaN if not found
        else:
            print("Warning: Cannot check Balance Sheet for shares (BS missing or empty).")
            shares_source += ", BS Unavailable"
            shares_outstanding = 0
        # Ensure shares is float/int, default to 0 if still invalid
        shares_outstanding = float(shares_outstanding) if np.isfinite(shares_outstanding) else 0.0

    # Get Beta - Try fetching dynamically, prompt user if fails
    beta_input = safe_get(info, ['beta'], 'Info', default=None) # Use None default to distinguish missing from zero beta
    if beta_input is not None and np.isfinite(beta_input):
        input_sources['Beta'] = 'Dynamic Fetch'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))
    else:
        print(f"⚠️ WARNING: Beta not available or invalid from dynamic fetch ({beta_input}).")
        # Ask user for Beta, allowing blank for default 1.0
        beta_user = get_validated_float_input(f"Enter Beta (β) for {ticker_symbol} (leave blank for 1.0):", default_value="1.0", allow_blank=True, min_val=-1.0, max_val=4.0)
        # beta_user will be None if blank, or the float value entered
        beta_input = float(beta_user) if beta_user is not None else 1.0 # Use 1.0 if user left blank
        info['beta'] = beta_input # Store the used beta back into info dict
        input_sources['Beta'] = 'User Input' if beta_user is not None else 'Default (1.0)'
        display(Markdown(f"**Beta (β):** {beta_input:.2f} (`Source: {input_sources['Beta']}`)"))

    # Market Cap - Try fetching, fallback to calculation
    market_cap = safe_get(info, ['marketCap'], 'Info', default=0)
    mcap_source = "Info"
    if (not np.isfinite(market_cap) or market_cap <= 0):
         mcap_source = "Info Invalid/Missing"
         if shares_outstanding > 0 and np.isfinite(current_price) and current_price > 0:
            market_cap = shares_outstanding * current_price
            info['marketCap'] = market_cap # Store calculated market cap back
            mcap_source = "Calculated (Price * Shares)"
            print(f"Info: Market Cap calculated from price and shares.")
         else:
            print(f"⚠️ WARNING: Market Cap N/A (cannot fetch or calculate).")
            market_cap = np.nan # Ensure it's NaN if invalid
            mcap_source = "N/A (Fetch & Calc Failed)"
    elif market_cap <=0: # Handle case where info gives 0 market cap
         market_cap = np.nan
         mcap_source = "Info Zero/Invalid"
         print(f"⚠️ WARNING: Market Cap from info is zero or invalid.")


    # Other Info
    sector = info.get("sector", "N/A"); industry = info.get("industry", "N/A"); summary = info.get("longBusinessSummary", "N/A.")

    # --- Display Summary ---
    display(Markdown(f"--- **Company Overview** ---"))
    display(Markdown(f"**Company:** {company_name}"))
    display(Markdown(f"**Sector:** {sector} | **Industry:** {industry}"))
    display(Markdown(f"**Currency:** {currency_code} (Using Symbol: {currency_symbol})")) # Show used symbol
    display(Markdown(f"**Current Price:** {format_currency(current_price)} (`Source: {price_source}`)"))
    display(Markdown(f"**Market Cap:** {format_currency(market_cap)} (`Source: {mcap_source}`)"))
    display(Markdown(f"**Shares Outstanding:** {shares_outstanding:,.0f} (`Source: {shares_source}`)"))
    if len(summary) > 5 and summary != "N/A.": display(Markdown(f"**Business Summary:** {summary[:300]}...")) # Show truncated summary


except Exception as e:
    print(f"❌ CRITICAL ERROR during data acquisition for {ticker_symbol}: {e}")
    print(traceback.format_exc())
    sys.exit("Halting execution due to data fetch error.")

# --- Final Sanity Checks Before Proceeding ---
if not isinstance(shares_outstanding,(int,float)) or shares_outstanding <= 0:
    print("CRITICAL: Shares outstanding is still invalid after fetch and fallbacks.")
    # Force user input as last resort
    shares_outstanding = get_validated_float_input(f"CRITICAL: Shares invalid ({shares_outstanding}). Enter Shares Outstanding:", min_val=1)
    shares_outstanding = float(shares_outstanding)
    input_sources['Shares'] = 'User Input (Critical Override)'
    print(f"Using user-provided shares: {shares_outstanding:,.0f}")

if beta_input is None or not np.isfinite(beta_input):
    print("CRITICAL: Beta is invalid after fetch and user prompt. Defaulting to 1.0.")
    beta_input = 1.0
    input_sources['Beta'] = 'Default (1.0) [Critical Override]'

if not np.isfinite(market_cap):
    print(f"⚠️ WARNING: Market Cap remains invalid. EV/EBITDA and WACC weights might be affected.")
    market_cap = np.nan # Ensure it's NaN

if not np.isfinite(current_price):
    print(f"⚠️ WARNING: Current Price remains invalid. P/E, P/B ratios and comparison will be affected.")
    current_price = np.nan


# %% [markdown]
# ## 5. WACC Calculation (Syntax Fix Applied V21.6)
# (Section Unchanged - Uses updated helpers, Syntax fix already present)

# %%
# WACC Function Definition
# Accepts already fetched data and user inputs
def calculate_wacc(stock_info_dict, financials_df, balance_sheet_df, user_rf, user_mrp, user_beta, input_sources_dict, currency_sym):
    # Declare global only if you intend to modify a global variable *from within* the function
    # 'user_credit_spread' is used but potentially set outside if needed, so access it directly
    global user_credit_spread # Needed to potentially *set* the global var if prompted inside

    display(Markdown("### Calculating Weighted Average Cost of Capital (WACC)..."))
    components = {} # To store parts of the WACC calculation
    calculation_log = [] # To store step-by-step process for display
    calculation_log.append(f"Input Sources -> Rf: {input_sources_dict.get('Rf','?')}, MRP: {input_sources_dict.get('MRP','?')}, Beta: {input_sources_dict.get('Beta','?')}")

    # --- Cost of Equity (Ke) ---
    beta_val = user_beta
    # Check for non-positive Beta, log warning but proceed (CAPM still works mathematically)
    # *** SYNTAX FIX ALREADY APPLIED: if statement on its own line ***
    if beta_val <= 0:
        calculation_log.append(f"⚠️ Warning: Beta is non-positive ({beta_val:.2f}). Cost of Equity might be low or negative.")
    components['beta'] = beta_val
    components['risk_free_rate'] = user_rf
    components['market_risk_premium'] = user_mrp

    cost_of_equity = user_rf + beta_val * user_mrp
    # Check if Cost of Equity is within a very broad reasonable range
    if not (0 < cost_of_equity < 0.50): # Check if Ke is between 0% and 50%
        calculation_log.append(f"⚠️ Warning: Calculated Cost of Equity ({format_percent(cost_of_equity)}) seems unusual.")
    components['cost_of_equity'] = cost_of_equity
    calculation_log.append(f"Inputs -> Rf: {format_percent(user_rf)}, Beta: {beta_val:.2f}, MRP: {format_percent(user_mrp)}")
    calculation_log.append(f"↳ Cost of Equity (Ke = Rf + Beta * MRP): {format_percent(cost_of_equity)}")

    # --- Cost of Debt (Kd) & Tax Rate ---
    cost_of_debt_pre_tax = np.nan
    tax_rate = np.nan
    after_tax_cost_of_debt = np.nan
    total_debt = 0.0
    debt_calculation_method = "N/A"
    tax_rate_source = "N/A"

    # Check if financial statements are available for calculations
    statements_available = (financials_df is not None and not financials_df.empty and
                            balance_sheet_df is not None and not balance_sheet_df.empty)

    if statements_available:
        latest_bs = balance_sheet_df.iloc[:, 0] # Most recent column
        latest_fin = financials_df.iloc[:, 0] # Most recent column

        # Get Debt from Balance Sheet
        debt_keys_long = ['Total Debt Non Current','Long Term Debt', 'Total Non Current Liabilities Net Minority Interest'] # Add more potential keys
        debt_keys_short = ['Total Debt Current','Current Debt', 'Total Current Liabilities'] # Add more potential keys
        long_term_debt = safe_get(latest_bs, debt_keys_long, 'BS(Long Debt)', 0)
        short_term_debt = safe_get(latest_bs, debt_keys_short, 'BS(Short Debt)', 0)

        # Ensure they are floats, default to 0 if not finite
        long_term_debt = float(long_term_debt) if np.isfinite(long_term_debt) else 0.0
        short_term_debt = float(short_term_debt) if np.isfinite(short_term_debt) else 0.0
        total_debt = long_term_debt + short_term_debt
        components['debt_value_used'] = total_debt # Store the debt value used
        calculation_log.append(f"Debt from BS -> LT Debt: {format_currency(long_term_debt)}, ST Debt: {format_currency(short_term_debt)} -> Total Debt: {format_currency(total_debt)}")

        # Get Interest Expense from Income Statement
        interest_keys = ['Interest Expense','Interest Expense Non Operating', 'Interest Expense Net']
        interest_expense = abs(safe_get(latest_fin, interest_keys, 'IS(Interest)', 0)) # Use absolute value
        interest_expense = float(interest_expense) if np.isfinite(interest_expense) else 0.0
        calculation_log.append(f"Interest from IS -> Interest Expense: {format_currency(interest_expense)}")

        # Try calculating Implied Cost of Debt (Kd)
        implied_rate_calculated = False
        if total_debt > 1000 and interest_expense > 0: # Avoid division by zero or tiny debt values
             implied_rate = interest_expense / total_debt
             implied_rate_calculated = True
             calculation_log.append(f"Implied Pre-tax Kd (Interest / Debt) = {format_percent(implied_rate)}")
             # Check if implied rate is within reasonable bounds
             if MIN_COST_OF_DEBT <= implied_rate <= MAX_COST_OF_DEBT:
                 cost_of_debt_pre_tax = implied_rate
                 debt_calculation_method = "Implied (Interest/Debt)"
                 calculation_log.append(f"↳ Using Implied Kd: {format_percent(cost_of_debt_pre_tax)}")
             else:
                 calculation_log.append(f"⚠️ Implied Kd ({format_percent(implied_rate)}) out of bounds [{format_percent(MIN_COST_OF_DEBT)}-{format_percent(MAX_COST_OF_DEBT)}]. Will use Rf + Spread method.")
                 implied_rate_calculated = False # Fallback to default method
        elif total_debt > 1000: # Has debt but no interest expense
             calculation_log.append(f"ℹ️ Cannot use implied Kd method (Interest Expense is zero). Will use Rf + Spread method.")
        else: # No significant debt
             calculation_log.append(f"ℹ️ Cannot use implied Kd method (Total Debt is zero or negligible).")

        # Default Kd Method (Rf + Credit Spread) if implied failed or no debt
        if not implied_rate_calculated:
             if total_debt > 1000: # Company has debt, need spread
                 debt_calculation_method = "Default (Rf + Spread)"
                 # Check if credit spread was already provided (e.g., from a previous run or input)
                 if user_credit_spread is None:
                     spread_prompt = f"Implied Kd calculation failed or not applicable. Enter Credit Spread % for {ticker_symbol} over Rf ({format_percent(user_rf)})"
                     # Get input as percentage
                     user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", allow_blank=False, min_val=0.1, max_val=15.0, is_percentage=True)
                     user_credit_spread = user_credit_spread_pct / 100.0 # Convert to rate
                     input_sources_dict['Credit Spread'] = 'User Input (for Default Kd)'
                 else:
                     input_sources_dict['Credit Spread'] = 'User Input (Reused)' # Indicate it was already set

                 cost_of_debt_pre_tax = user_rf + user_credit_spread
                 original_kd = cost_of_debt_pre_tax # Store before bounding
                 # Apply bounds to the calculated Kd
                 cost_of_debt_pre_tax = min(max(cost_of_debt_pre_tax, MIN_COST_OF_DEBT), MAX_COST_OF_DEBT)
                 calculation_log.append(f"Default Pre-tax Kd = Rf + Spread = {format_percent(user_rf)} + {format_percent(user_credit_spread)} = {format_percent(original_kd)}")
                 if abs(cost_of_debt_pre_tax - original_kd) > 1e-6: # Check if bounding was applied
                      calculation_log.append(f"↳ Bounded Default Kd: {format_percent(cost_of_debt_pre_tax)} (within [{format_percent(MIN_COST_OF_DEBT)}-{format_percent(MAX_COST_OF_DEBT)}])")
                 else:
                      calculation_log.append(f"↳ Using Default Kd: {format_percent(cost_of_debt_pre_tax)}")
                 calculation_log.append(f"   (Credit Spread Source: {input_sources_dict['Credit Spread']})")

             else: # Company has no significant debt
                 debt_calculation_method = "Zero Debt"
                 cost_of_debt_pre_tax = 0.0 # Cost of debt is 0 if there's no debt
                 input_sources_dict['Credit Spread'] = 'N/A (Zero Debt)'
                 calculation_log.append(f"ℹ️ Pre-tax Kd set to 0% (Zero or negligible debt).")

        # Calculate Effective Tax Rate
        tax_exp_keys = ['Tax Provision','Income Tax Expense']
        pretax_inc_keys = ['Pretax Income','EBT', 'Earnings Before Tax']
        income_tax_expense = safe_get(latest_fin, tax_exp_keys, 'IS(Tax Exp)', 0)
        pretax_income = safe_get(latest_fin, pretax_inc_keys, 'IS(Pretax Inc)', 0)
        income_tax_expense = float(income_tax_expense) if np.isfinite(income_tax_expense) else 0.0
        pretax_income = float(pretax_income) if np.isfinite(pretax_income) else 0.0
        calculation_log.append(f"Tax Rate Calc -> Tax Expense: {format_currency(income_tax_expense)}, Pre-tax Income: {format_currency(pretax_income)}")

        # Handle cases where pretax income is zero or negative, or tax expense is negative
        if pretax_income <= 1 and income_tax_expense > 0: # Tax paid despite low/no pretax income
             # Try using EBIT - Interest as a proxy for Pretax Income if EBIT is available
             ebit_val = safe_get(latest_fin, ["EBIT"], 'IS(EBIT for Tax Proxy)', default=0)
             ebit_val = float(ebit_val) if np.isfinite(ebit_val) else 0.0
             if ebit_val > 1 and np.isfinite(interest_expense):
                 pretax_proxy = ebit_val - interest_expense
                 if pretax_proxy > 1:
                     pretax_income = pretax_proxy
                     calculation_log.append(f"ℹ️ Pretax Income was low/zero, using EBIT - Interest proxy: {format_currency(pretax_income)}")
                 else: calculation_log.append(f"⚠️ Cannot calculate meaningful tax rate (Pretax Income and Proxy are too low).")
             else: calculation_log.append(f"⚠️ Cannot calculate meaningful tax rate (Pretax Income low, EBIT unavailable/low).")

        if pretax_income > 1 and income_tax_expense >= 0: # Can calculate effective rate
            calc_tax_rate = income_tax_expense / pretax_income
            calculation_log.append(f"Calculated Effective Tax Rate = {format_percent(calc_tax_rate)}")
            # Apply bounds to the calculated tax rate
            original_tax_rate = calc_tax_rate
            tax_rate = min(max(calc_tax_rate, MIN_TAX_RATE), MAX_TAX_RATE)
            tax_rate_source = "Derived (Tax Exp / Pretax Inc)"
            if abs(tax_rate - original_tax_rate) > 0.001: # Check if bounding was applied
                tax_rate_source += f" [Bounded to {format_percent(MIN_TAX_RATE)}-{format_percent(MAX_TAX_RATE)}]"
                calculation_log.append(f"↳ Bounded Effective Tax Rate: {format_percent(tax_rate)}")
            else:
                 calculation_log.append(f"↳ Using Effective Tax Rate: {format_percent(tax_rate)}")
        else: # Cannot derive tax rate
            tax_rate = DEFAULT_TAX_RATE # Use default tax rate
            tax_rate_source = f"Default ({format_percent(DEFAULT_TAX_RATE)}) [Derived Failed]"
            calculation_log.append(f"⚠️ Could not derive tax rate. Using default: {format_percent(tax_rate)}")

    else: # Financial statements missing
        calculation_log.append("⚠️ Financial statements missing. Using defaults for Cost of Debt and Tax Rate.")
        debt_calculation_method = "Default (No Financials)"
        tax_rate_source = f"Default ({format_percent(DEFAULT_TAX_RATE)}) [No Financials]"

        # Need to ask for credit spread if statements are missing (and spread not already set)
        if user_credit_spread is None:
             spread_prompt=f"Financials missing. Enter Credit Spread % for {ticker_symbol} over Rf ({format_percent(user_rf)})"
             user_credit_spread_pct = get_validated_float_input(spread_prompt, default_value=f"{DEFAULT_CREDIT_SPREAD*100:.1f}", min_val=0.1, max_val=15.0, is_percentage=True)
             user_credit_spread = user_credit_spread_pct / 100.0
             input_sources_dict['Credit Spread'] = 'User Input (No Financials)'
        else:
             input_sources_dict['Credit Spread'] = 'User Input (Reused)'

        cost_of_debt_pre_tax = user_rf + user_credit_spread
        original_kd = cost_of_debt_pre_tax
        cost_of_debt_pre_tax = min(max(cost_of_debt_pre_tax, MIN_COST_OF_DEBT), MAX_COST_OF_DEBT) # Apply bounds
        tax_rate = DEFAULT_TAX_RATE
        total_debt = 0 # Assume zero debt if no financials
        components['debt_value_used'] = 0 # Store zero debt used

        calculation_log.append(f"Default Pre-tax Kd = Rf + Spread = {format_percent(original_kd)}")
        if abs(cost_of_debt_pre_tax - original_kd) > 1e-6: calculation_log.append(f"↳ Bounded Default Kd: {format_percent(cost_of_debt_pre_tax)}")
        else: calculation_log.append(f"↳ Using Default Kd: {format_percent(cost_of_debt_pre_tax)}")
        calculation_log.append(f"Using Default Tax Rate: {format_percent(tax_rate)}")

    # Store Kd and Tax Rate components
    components['cost_of_debt'] = cost_of_debt_pre_tax if np.isfinite(cost_of_debt_pre_tax) else 0.0
    components['debt_calc_method'] = debt_calculation_method
    components['tax_rate'] = tax_rate if np.isfinite(tax_rate) else DEFAULT_TAX_RATE # Fallback if NaN
    components['tax_rate_source'] = tax_rate_source

    # Calculate After-Tax Cost of Debt
    if np.isfinite(components['cost_of_debt']) and np.isfinite(components['tax_rate']):
        after_tax_cost_of_debt = components['cost_of_debt'] * (1 - components['tax_rate'])
    else:
        calculation_log.append(f"⚠️ Cannot calculate after-tax cost of debt (Kd or Tax Rate invalid).")
        after_tax_cost_of_debt = np.nan # Keep as NaN if calculation fails
    components['after_tax_cost_of_debt'] = after_tax_cost_of_debt
    if np.isfinite(after_tax_cost_of_debt) and components['cost_of_debt'] > 0: # Only log if debt exists
        calculation_log.append(f"↳ Cost of Debt (After-tax Kd * (1-T)): {format_percent(after_tax_cost_of_debt)}")
    elif components['cost_of_debt'] == 0:
        calculation_log.append(f"↳ Cost of Debt (After-tax): 0.00% (Due to zero pre-tax Kd)")


    # --- Market Value Weights (Equity & Debt) ---
    # Use Market Cap for Equity value if available and valid
    market_cap_val = safe_get(stock_info_dict, ['marketCap'], 'Info', default=np.nan)
    equity_value_for_wacc = np.nan
    equity_source = "N/A"

    if np.isfinite(market_cap_val) and market_cap_val > 0:
        equity_value_for_wacc = market_cap_val
        equity_source = "Market Cap (from Info/Calculated)"
    else:
        calculation_log.append(f"⚠️ Market Cap invalid ({format_currency(market_cap_val)}). Attempting fallback to Book Equity.")
        equity_source = "Market Cap Invalid"
        # Fallback to Book Value of Equity if Market Cap invalid
        if balance_sheet_df is not None and not balance_sheet_df.empty:
            equity_keys = ['Total Stockholder Equity', 'Common Stock Equity', 'Stockholders Equity']
            book_equity_val = safe_get(balance_sheet_df.iloc[:, 0], equity_keys, 'BS(Book Eq)', default=np.nan)
            if np.isfinite(book_equity_val) and book_equity_val > 1.0: # Check if book equity is positive
                equity_value_for_wacc = book_equity_val
                equity_source = "Book Equity (BS Fallback)"
                calculation_log.append(f"↳ Using Book Value of Equity as proxy: {format_currency(equity_value_for_wacc)}")
            else:
                equity_source += ", Book Equity Invalid/Zero"
                calculation_log.append(f"⚠️ Book Equity from BS also invalid ({format_currency(book_equity_val)}).")
        else:
             equity_source += ", BS Unavailable"
             calculation_log.append(f"⚠️ Cannot use Book Equity (Balance Sheet unavailable).")

        # Ultimate fallback if both market and book equity fail
        if not np.isfinite(equity_value_for_wacc):
            equity_value_for_wacc = 1.0 # Use a placeholder to avoid division by zero, but log warning
            equity_source = "Placeholder (1.0) [All Sources Failed]"
            calculation_log.append("⚠️ CRITICAL: Using placeholder Equity value (1.0) for WACC weights. Result unreliable.")

    components['equity_value_used'] = equity_value_for_wacc
    components['equity_source'] = equity_source

    # Calculate Weights
    equity_weight = np.nan; debt_weight = np.nan; weights_source = "N/A"
    # Use total_debt calculated earlier
    total_capital = equity_value_for_wacc + total_debt

    if np.isfinite(total_capital) and total_capital > 1.0: # Ensure total capital is positive
        equity_weight = equity_value_for_wacc / total_capital
        debt_weight = total_debt / total_capital
        weights_source = "Calculated (E/(E+D), D/(E+D))"
        # Check if weights sum close to 1, normalize if slightly off due to floats
        total_weight = equity_weight + debt_weight
        if abs(total_weight - 1.0) > 1e-6:
             calculation_log.append(f"ℹ️ Weights sum to {total_weight:.4f}. Normalizing.")
             equity_weight /= total_weight
             debt_weight /= total_weight
    else:
        calculation_log.append(f"❌ ERROR: Total Capital calculation invalid ({format_currency(total_capital)}). Using default weights (70% E / 30% D). WACC unreliable.")
        equity_weight = 0.7
        debt_weight = 0.3
        weights_source = "Default Weights (Invalid Capital)"

    components['equity_weight'] = equity_weight
    components['debt_weight'] = debt_weight
    components['weights_source'] = weights_source

    calculation_log.append(f"Capital Values -> Equity ({components['equity_source']}): {format_currency(components['equity_value_used'])}, Debt ({components['debt_calc_method']}): {format_currency(total_debt)}")
    calculation_log.append(f"↳ Weights -> E/V: {format_percent(equity_weight)}, D/V: {format_percent(debt_weight)} ({weights_source})")

    # --- Final WACC Calculation ---
    wacc = np.nan
    # Check if all required components are finite numbers
    if all(map(np.isfinite, [equity_weight, cost_of_equity, debt_weight, after_tax_cost_of_debt])):
        wacc = (equity_weight * cost_of_equity) + (debt_weight * after_tax_cost_of_debt)
        original_wacc = wacc # Store before bounding
        # Apply WACC bounds
        wacc = min(max(wacc, MIN_WACC), MAX_WACC)
        bound_note = ""
        if abs(wacc - original_wacc) > 0.0001 :
            bound_note = f" (Bounded to {format_percent(MIN_WACC)}-{format_percent(MAX_WACC)})"
            calculation_log.append(f"ℹ️ Calculated WACC ({format_percent(original_wacc)}) was outside bounds. Applied bounds.")
        calculation_log.append(f"▶️ WACC = (E/V * Ke) + (D/V * Kd * (1-T)) = {format_percent(wacc)}{bound_note}")
    else:
        calculation_log.append(f"❌ ERROR: Cannot calculate final WACC. One or more components invalid:")
        if not np.isfinite(equity_weight): calculation_log.append("   - Equity Weight Invalid")
        if not np.isfinite(cost_of_equity): calculation_log.append("   - Cost of Equity Invalid")
        if not np.isfinite(debt_weight): calculation_log.append("   - Debt Weight Invalid")
        if not np.isfinite(after_tax_cost_of_debt): calculation_log.append("   - After-Tax Cost of Debt Invalid")
        wacc = np.nan # Ensure WACC is NaN if calculation fails

    components['final_wacc'] = wacc # Store final WACC (potentially bounded or NaN)
    display(Markdown("```\n" + "\n".join(calculation_log) + "\n```")) # Display calculation log

    # --- WACC Composition Pie Chart ---
    if np.isfinite(wacc) and wacc > 0 and np.isfinite(equity_weight) and np.isfinite(debt_weight):
        equity_contrib = equity_weight * cost_of_equity
        debt_contrib = debt_weight * after_tax_cost_of_debt

        # Ensure contributions are valid and roughly sum to WACC before plotting
        if np.isfinite(equity_contrib) and np.isfinite(debt_contrib) and abs(equity_contrib + debt_contrib - wacc) < 0.01: # Allow small tolerance
           labels = [f'Equity Contribution ({format_percent(equity_contrib)})', f'Debt Contribution ({format_percent(debt_contrib)})']
           values = [equity_contrib, debt_contrib]
           fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4,
                                        marker_colors=px.colors.qualitative.Pastel,
                                        textinfo='percent', sort=False,
                                        hoverinfo='label+value')])
           fig.update_layout(title_text=f'WACC Composition (Total: {format_percent(wacc)})', title_x=0.5,
                             showlegend=True, legend_title_text='Components', height=350,
                             margin=dict(t=60, b=30, l=30, r=30))
           fig.show()
        else:
            display(Markdown(f"*(WACC contribution calculation invalid. Cannot generate composition chart.)*"))
            # print(f"Debug: Eq Contrib={equity_contrib}, Debt Contrib={debt_contrib}, WACC={wacc}") # Debugging line
    elif not np.isfinite(wacc):
         display(Markdown(f"*(WACC calculation failed. No composition chart.)*"))
    else: # WACC is 0 or negative
         display(Markdown(f"*(WACC is zero or negative ({format_percent(wacc)}). No composition chart.)*"))

    return wacc, components # Return the final WACC and the components dictionary

# --- Execute WACC Calculation ---
# Pass the required data fetched earlier and user inputs
# currency_symbol is already forced to '₹'
discount_rate, wacc_components = calculate_wacc(
    stock_data['info'],
    stock_data.get('financials'), # Use .get to handle potentially missing statements gracefully
    stock_data.get('balance_sheet'),
    risk_free_rate_input,
    market_risk_premium_input,
    beta_input,
    input_sources, # Pass the dictionary tracking input sources
    currency_symbol # Pass the globally determined currency symbol ('₹')
)


# %% [markdown]
# ## 6. FCFF Method Selection (Informed Choice - V21.6 Fix)
# (Section Unchanged - Uses updated helpers)

# %%
# --- Pre-calculate Averages and Get User Choice ---
selected_fcff_method = None
precalculated_averages = {}
fcff_choice_log = []

print("\n--- FCFF Calculation Method Selection ---")
fcff_choice_log.append("--- Pre-calculating 4yr Average FCFF for Each Method ---")

# Use global currency symbol determined earlier ('₹')
global_currency_symbol = currency_symbol

# Get the tax rate calculated during WACC, with fallback
wacc_tax_rate = wacc_components.get('tax_rate', DEFAULT_TAX_RATE) if isinstance(wacc_components, dict) else DEFAULT_TAX_RATE
if not np.isfinite(wacc_tax_rate):
    fcff_choice_log.append(f"  ⚠️ Warning: Invalid tax rate from WACC calculation ({wacc_tax_rate}). Using default {format_percent(DEFAULT_TAX_RATE)} for FCFF pre-calculation.")
    wacc_tax_rate = DEFAULT_TAX_RATE

fcff_choice_log.append(f"  Using Tax Rate for pre-calc: {format_percent(wacc_tax_rate)} (Source: {wacc_components.get('tax_rate_source','Default Fallback')})")


# Check if all necessary dataframes are available and tax rate is valid
data_ok_for_fcff = (stock_data.get('financials') is not None and not stock_data['financials'].empty and
                    stock_data.get('balance_sheet') is not None and not stock_data['balance_sheet'].empty and
                    stock_data.get('cashflow') is not None and not stock_data['cashflow'].empty and
                    np.isfinite(wacc_tax_rate))

if data_ok_for_fcff:
    num_hist_years_fcff = 4 # Target number of years for averaging
    # Determine the actual number of years available across all statements
    num_statement_years_fcff = 0
    try:
        num_statement_years_fcff = min(stock_data['financials'].shape[1],
                                     stock_data['balance_sheet'].shape[1],
                                     stock_data['cashflow'].shape[1])
    except Exception as e:
         fcff_choice_log.append(f"  ❌ ERROR determining available statement years: {e}")

    # Years to use for averaging is the minimum of target and available
    years_to_avg_fcff = min(num_statement_years_fcff, num_hist_years_fcff)

    if years_to_avg_fcff < 1:
        fcff_choice_log.append(f"  ❌ ERROR: Insufficient historical data available ({years_to_avg_fcff} years) to calculate average FCFF.")
        data_ok_for_fcff = False # Mark as not OK if we can't average even 1 year
    else:
        fcff_choice_log.append(f"  Calculating averages over the last {years_to_avg_fcff} available year(s).")
        for method_key, method_info in FCFF_METHODS_DEFINITION.items():
            fcff_choice_log.append(f"\n  Avg for Method {method_key}: {method_info['name']}...")
            temp_hist_fcff = [] # Store FCFF for each year for this method
            detailed_calc_logs = [] # Store logs for each year's calculation

            for i in range(years_to_avg_fcff):
                 year_label = f"T-{i}" # e.g., T-0 (most recent), T-1, etc.
                 # Ensure we don't go out of bounds for any statement
                 if i >= stock_data['financials'].shape[1] or \
                    i >= stock_data['balance_sheet'].shape[1] or \
                    i >= stock_data['cashflow'].shape[1]:
                     detailed_calc_logs.append(f"    {year_label}: Skipped (Insufficient data for this year)")
                     temp_hist_fcff.append(np.nan)
                     continue

                 # Get data for the specific year (column i)
                 fin_c = stock_data['financials'].iloc[:, i]
                 bs_c = stock_data['balance_sheet'].iloc[:, i]
                 cf_c = stock_data['cashflow'].iloc[:, i]
                 # Get previous year's balance sheet (column i+1) if it exists
                 bs_p = stock_data['balance_sheet'].iloc[:, i+1] if (i+1) < stock_data['balance_sheet'].shape[1] else None

                 # Calculate FCFF for this specific year and method
                 temp_calc_log_for_year = [] # Log for this specific year/method calc
                 fcff_yr_val = calculate_fcff_single_method_single_year(
                     method_key, fin_c, bs_c, cf_c, bs_p, wacc_tax_rate, temp_calc_log_for_year
                 )
                 temp_hist_fcff.append(fcff_yr_val)
                 # Store the detailed log for this year
                 detailed_calc_logs.append(f"--- {year_label} Calculation Log (Method {method_key}) ---")
                 detailed_calc_logs.extend([f"    {line}" for line in temp_calc_log_for_year]) # Indent sub-log
                 detailed_calc_logs.append(f"    {year_label} FCFF Result: {format_currency(fcff_yr_val)}")
                 detailed_calc_logs.append("-" * 20) # Separator


            # Calculate the average FCFF for this method, ignoring NaNs
            valid_fcff_values = [f for f in temp_hist_fcff if np.isfinite(f)]
            valid_count = len(valid_fcff_values)
            avg_fcff = np.mean(valid_fcff_values) if valid_count > 0 else np.nan # Use mean, could use median
            precalculated_averages[method_key] = avg_fcff

            # Log the result for this method
            fcff_choice_log.append(f"    Individual Year FCFFs ({years_to_avg_fcff} yrs): {[format_currency(f) for f in temp_hist_fcff]}")
            if valid_count > 0:
                 fcff_choice_log.append(f"    Resulting {valid_count}-year Avg FCFF: {format_currency(avg_fcff)}") # format_currency handles symbol/precision
            else:
                 fcff_choice_log.append(f"    Resulting Avg FCFF: N/A (No valid yearly calculations)")
            # Optionally display detailed logs (can be very verbose)
            # fcff_choice_log.extend(detailed_calc_logs)


    # --- Present Choices to User ---
    print("\nPlease choose the FCFF calculation method based on the calculated historical averages:")
    valid_method_choices = [] # Store keys of methods with positive average FCFF
    display_options = [] # Store formatted strings for display

    for method_key, method_info in FCFF_METHODS_DEFINITION.items():
        avg_val = precalculated_averages.get(method_key, np.nan)
        avg_display = "N/A (Calc Failed/No Data)"; is_positive_and_valid = False

        if np.isfinite(avg_val):
            # Use updated format_currency (handles symbol automatically)
            avg_display = format_currency(avg_val) # Default precision=2
            if avg_val > 1e-9: # Check if strictly positive (allowing for float inaccuracy)
                 is_positive_and_valid = True
                 avg_display += " ✅ (Positive Avg - Usable for DCF)"
            else:
                 avg_display += " ⚠️ (Zero or Negative Avg - Cannot use for DCF basis)"
        elif method_key not in precalculated_averages:
            avg_display = "N/A (Calculation Not Attempted - Likely Missing Data)"

        # Build display string
        option_str = f"\n  {method_key}. {method_info['name']}\n"
        option_str += f"     Formula: {method_info['formula']}\n"
        option_str += f"     {years_to_avg_fcff}yr Avg FCFF: {avg_display}"
        display_options.append(option_str)

        if is_positive_and_valid:
            valid_method_choices.append(method_key) # Add to list of valid choices

    # Print all options
    print("--- Available Methods and Average FCFF ---")
    for option in display_options:
        print(option)
    print("------------------------------------------")


    # --- Get User Selection ---
    if not valid_method_choices:
         print("\n❌ CRITICAL: No FCFF methods resulted in a positive average FCFF over the available period.")
         print("   Cannot proceed with DCF valuation using the FCFF method as it requires a positive starting basis.")
         selected_fcff_method = None
         input_sources['FCFF Method'] = "None (No Positive Avg FCFF Found)"
         # Optional: Exit here if FCFF is absolutely required
         # sys.exit("Halting: No positive average FCFF basis found.")
    else:
        prompt_text = f"Enter the number ({', '.join(map(str, valid_method_choices))}) for the FCFF method you want to use for the DCF valuation:"
        while selected_fcff_method is None:
            try:
                choice_str = input(f"{prompt_text}: ").strip()
                if not choice_str: # Handle empty input
                    print("❌ Please enter a valid method number.")
                    continue
                method_idx = int(choice_str)
                if method_idx in valid_method_choices:
                    selected_fcff_method = method_idx
                    method_name_chosen = FCFF_METHODS_DEFINITION[selected_fcff_method]['name']
                    chosen_avg = precalculated_averages[selected_fcff_method]
                    # Use updated format_currency
                    display(Markdown(f"**Selected FCFF Method:** `{method_idx} - {method_name_chosen}`\n"
                                     f"  *Using 4yr Avg Basis: {format_currency(chosen_avg)}*"))
                    input_sources['FCFF Method'] = f"User Choice ({method_name_chosen} - Method {method_idx})"
                else:
                    print(f"❌ Invalid choice. Please enter one of the method numbers with a positive average: {valid_method_choices}.")
            except ValueError:
                print("❌ Invalid input. Please enter a number.")
            except Exception as e:
                print(f"An unexpected error occurred during method selection: {e}")
                sys.exit("Halting due to error during FCFF method selection.")

else: # Initial data check failed
    print("\n❌ ERROR: Cannot proceed with FCFF method selection due to missing financial data or invalid tax rate.")
    fcff_choice_log.append("  ❌ ERROR: Prerequisite data missing (Financials/BS/CF or Tax Rate).")
    selected_fcff_method = None
    input_sources['FCFF Method'] = "None (Prerequisite Data Missing)"

# Optionally display the log of the pre-calculation process
# display(Markdown("```\n" + "\n".join(fcff_choice_log) + "\n```"))


# %% [markdown]
# ## 7. FCFF Calculation & Projection (V21.6 - No Stock CAGR / Formatting)
# (Section Unchanged - Uses updated helpers)

# %%
# FCFF Function Definition (V21.6 - No Stock CAGR, Uses Selected Method Basis)
def calculate_and_project_fcff_user_selected_method_v21_6(
                                   user_selected_method_idx, # The method chosen by the user (e.g., 0, 1, 2...)
                                   precalculated_avg_basis, # The positive avg FCFF calculated for the chosen method
                                   financials_df, balance_sheet_df, cashflow_df, # Historical data
                                   wacc_comps, # WACC components dict (for tax rate)
                                   disc_rate, # The final WACC discount rate
                                   forecast_yrs, # Number of years to forecast (e.g., 3)
                                   final_user_rf, # User's risk-free rate (for terminal growth cap)
                                   default_init_growth, # Fallback initial growth rate
                                   currency_sym): # Currency symbol ('₹')

    method_info = FCFF_METHODS_DEFINITION.get(user_selected_method_idx, {"name": "Unknown Method"})
    method_name_display = method_info['name']
    display(Markdown(f"### Calculating & Projecting FCFF (Using Selected Method: {user_selected_method_idx} - {method_name_display})"))
    projection_log = []
    projection_results = {
        'historical_fcffs_selected_method': {}, # Dict: {T-0: val, T-1: val, ...}
        'historical_revenues': {}, # Dict: {T-0: val, T-1: val, ...}
        'current_fcff_basis': precalculated_avg_basis, # The starting point for projection
        'fcff_basis_method': f"{method_name_display} (User Selected Method {user_selected_method_idx} - 4yr Avg)", # Description of basis
        'projected_fcffs': [], # List: [FCFF_Y1, FCFF_Y2, ...]
        'growth_rates': [], # List: [Growth_Y1, Growth_Y2, ...]
        'pv_fcffs': [], # List: [PV_FCFF_Y1, PV_FCFF_Y2, ...]
        'cumulative_pv_fcff': np.nan,
        'initial_growth_rate': np.nan, # The calculated initial growth rate used
        'growth_source': 'N/A', # How initial growth was determined
        'terminal_growth_rate': np.nan, # The calculated terminal growth rate (g)
        'terminal_value': np.nan,
        'pv_terminal_value': np.nan,
        'error': None # To store any fatal error message
    }

    # --- Pre-checks ---
    if not np.isfinite(disc_rate):
        projection_results['error']="Invalid WACC provided."; projection_log.append(f"❌ ERROR: WACC is not finite ({disc_rate}). Cannot proceed."); display(Markdown("```\n"+"\n".join(projection_log)+"\n```")); return projection_results
    if not np.isfinite(precalculated_avg_basis) or precalculated_avg_basis <= 1e-9:
        projection_results['error']=f"Invalid/non-positive FCFF basis ({format_currency(precalculated_avg_basis)}) for selected method {user_selected_method_idx}. Cannot proceed."; projection_log.append(f"❌ ERROR: {projection_results['error']}"); display(Markdown("```\n"+"\n".join(projection_log)+"\n```")); return projection_results

    # Get tax rate from WACC components, with fallback
    effective_tax_rate = wacc_components.get('tax_rate', DEFAULT_TAX_RATE) if isinstance(wacc_components, dict) else DEFAULT_TAX_RATE
    if not np.isfinite(effective_tax_rate): effective_tax_rate = DEFAULT_TAX_RATE
    tax_rate_src = wacc_components.get('tax_rate_source', '?') if isinstance(wacc_components, dict) else 'Default Fallback'
    projection_log.append(f"Using WACC/Discount Rate: {format_percent(disc_rate)}")
    projection_log.append(f"Using Effective Tax Rate: {format_percent(effective_tax_rate)} (Source: {tax_rate_src})")
    projection_log.append(f"FCFF Basis for Projection (T=0): {format_currency(precalculated_avg_basis)} (Source: {projection_results['fcff_basis_method']})")


    # --- PHASE 1: Re-Calculate Historical FCFF & Revenue using the SELECTED method ---
    # This is useful for calculating CAGR based *only* on the chosen method's results
    projection_log.append(f"\n--- Phase 1: Re-Calculating Historical Data (Using ONLY Method {user_selected_method_idx}) ---")
    num_hist_years_needed = 4 # Need T-0, T-1, T-2, T-3 for 3-year CAGR
    num_statement_years_avail = 0
    try: num_statement_years_avail = min(financials_df.shape[1], balance_sheet_df.shape[1], cashflow_df.shape[1])
    except: pass # Handle cases where dfs might be None initially (though pre-check should catch this)

    hist_fcffs_sel = [] # Store historical FCFF calculated with the selected method
    hist_revs = [] # Store corresponding historical revenues

    years_to_calc_hist = min(num_statement_years_avail, num_hist_years_needed)
    if years_to_calc_hist < num_hist_years_needed:
        projection_log.append(f"  ⚠️ Warning: Only {years_to_calc_hist} years of full statement data available. CAGR calculation might be affected.")

    for i in range(years_to_calc_hist):
        year_label = f"T-{i}"
        projection_log.append(f"\n--- Calculating Hist Year {year_label} (Method {user_selected_method_idx}) ---")
        # Data already checked for availability for this range
        fin = financials_df.iloc[:, i]; bs = balance_sheet_df.iloc[:, i]; cf = cashflow_df.iloc[:, i]
        bs_p = balance_sheet_df.iloc[:, i+1] if (i+1) < balance_sheet_df.shape[1] else None

        # Calculate FCFF for this year using the single selected method
        log_for_hist_calc = [] # Temporary log for this calculation
        fcff_val = calculate_fcff_single_method_single_year(user_selected_method_idx, fin, bs, cf, bs_p, effective_tax_rate, log_for_hist_calc)
        hist_fcffs_sel.append(fcff_val)
        projection_log.extend([f"  {line}" for line in log_for_hist_calc]) # Append indented log
        projection_log.append(f"  --> Hist FCFF ({year_label}, Method {user_selected_method_idx}): {format_currency(fcff_val)}")

        # Get revenue for the same year
        rev_val = safe_get(fin, ["Total Revenue", "Revenue"], "IS", default=np.nan)
        hist_revs.append(rev_val)
        projection_log.append(f"  --> Hist Revenue ({year_label}): {format_currency(rev_val)}")

    # Pad with NaNs if fewer than desired years were calculated
    while len(hist_fcffs_sel) < num_hist_years_needed: hist_fcffs_sel.append(np.nan)
    while len(hist_revs) < num_hist_years_needed: hist_revs.append(np.nan)

    # Reverse lists so T-0 is at index 0, T-1 at index 1, etc. (easier for CAGR)
    # Example: If calculated [FCFF_T0, FCFF_T1, FCFF_T2, FCFF_T3], they are already in correct order T-0, T-1,...
    projection_results['historical_fcffs_selected_method'] = {f"T-{i}": v for i, v in enumerate(hist_fcffs_sel)}
    projection_results['historical_revenues'] = {f"T-{i}": v for i, v in enumerate(hist_revs)}


    # --- PHASE 2: Estimate Initial Growth Rate (Based on Rev & Selected FCFF CAGR ONLY - V21.6) ---
    projection_log.append("\n--- Phase 2: Estimating Initial Growth (Geometric Mean of Revenue & Selected FCFF CAGRs) ---")
    init_growth = np.nan; growth_src = "N/A"; cagr_calc_details = {}; cagr_list_for_gm = []
    num_yrs_cagr = 3 # Use 3-year CAGR (requires 4 data points: T-0, T-1, T-2, T-3)

    # Revenue CAGR
    rev_cagr = np.nan
    projection_log.append(f"\nCalculating Revenue {num_yrs_cagr}yr CAGR...")
    if len(hist_revs) >= num_hist_years_needed: # Need at least 4 years of data
        rev_end = hist_revs[0] # T-0 value
        rev_start = hist_revs[num_yrs_cagr] # T-3 value
        projection_log.append(f"  Revenue Data: Start (T-{num_yrs_cagr}) = {format_currency(rev_start)}, End (T-0) = {format_currency(rev_end)}")
        # Check if start and end values are valid for CAGR calculation
        if pd.notna(rev_end) and math.isfinite(rev_end) and pd.notna(rev_start) and math.isfinite(rev_start) and rev_start > 1e-9:
            rev_cagr = calculate_cagr_modified(rev_start, rev_end, num_yrs_cagr)
            if pd.notna(rev_cagr) and math.isfinite(rev_cagr):
                cagr_calc_details['Revenue CAGR'] = rev_cagr
                cagr_list_for_gm.append(rev_cagr)
                projection_log.append(f"  ✅ Revenue {num_yrs_cagr}yr CAGR: {format_percent(rev_cagr)}")
            else: projection_log.append(f"  ❌ Revenue CAGR calculation failed (Result: {rev_cagr}).")
        else: projection_log.append(f"  ❌ Revenue CAGR calculation skipped (Invalid start/end values).")
    else: projection_log.append(f"  ❌ Revenue CAGR calculation skipped (Insufficient historical data: {len(hist_revs)} years).")

    # FCFF CAGR (Using ONLY the selected method's historical results)
    fcff_cagr = np.nan
    projection_log.append(f"\nCalculating FCFF {num_yrs_cagr}yr CAGR (Using Method {user_selected_method_idx} Results)...")
    if len(hist_fcffs_sel) >= num_hist_years_needed:
        fcff_end = hist_fcffs_sel[0] # T-0 value (selected method)
        fcff_start = hist_fcffs_sel[num_yrs_cagr] # T-3 value (selected method)
        projection_log.append(f"  FCFF Data (Method {user_selected_method_idx}): Start (T-{num_yrs_cagr}) = {format_currency(fcff_start)}, End (T-0) = {format_currency(fcff_end)}")
        # Check if start and end values are valid for CAGR calculation (must be positive for FCFF CAGR)
        if pd.notna(fcff_end) and math.isfinite(fcff_end) and fcff_end > 1e-9 and \
           pd.notna(fcff_start) and math.isfinite(fcff_start) and fcff_start > 1e-9:
            fcff_cagr = calculate_cagr_modified(fcff_start, fcff_end, num_yrs_cagr)
            if pd.notna(fcff_cagr) and math.isfinite(fcff_cagr):
                cagr_calc_details[f'FCFF (Method {user_selected_method_idx}) CAGR'] = fcff_cagr
                cagr_list_for_gm.append(fcff_cagr)
                projection_log.append(f"  ✅ FCFF {num_yrs_cagr}yr CAGR: {format_percent(fcff_cagr)}")
            else: projection_log.append(f"  ❌ FCFF CAGR calculation failed (Result: {fcff_cagr}).")
        else: projection_log.append(f"  ❌ FCFF CAGR calculation skipped (Start/end values non-positive or invalid).")
    else: projection_log.append(f"  ❌ FCFF CAGR calculation skipped (Insufficient historical data: {len(hist_fcffs_sel)} years).")

    # Calculate Geometric Mean of the valid calculated CAGRs
    projection_log.append("\nCalculating Geometric Mean of available valid CAGRs...")
    projection_log.append(f"  Input rates for GM: {[format_percent(r) for r in cagr_list_for_gm]}")
    if cagr_list_for_gm: # If we have at least one valid CAGR
        gm_growth = geometric_mean_modified(cagr_list_for_gm)
        if pd.notna(gm_growth) and math.isfinite(gm_growth):
            init_growth = gm_growth
            growth_src = f"Geometric Mean of {len(cagr_list_for_gm)} valid CAGR(s)"
            projection_log.append(f"  ▶️ Calculated Initial Growth (GM): {format_percent(init_growth)}")
        else:
            init_growth = default_init_growth # Fallback to default if GM fails
            growth_src = f"Default ({format_percent(default_init_growth)}) [GM calculation failed]"
            projection_log.append(f"  ⚠️ Geometric Mean calculation failed. Using default growth rate.")
    else: # No valid CAGRs calculated
        init_growth = default_init_growth # Fallback to default
        growth_src = f"Default ({format_percent(default_init_growth)}) [No valid CAGRs found]"
        projection_log.append(f"  ⚠️ No valid CAGRs found. Using default growth rate.")

    # Apply Bounds to the initial growth rate
    unbounded_growth = init_growth
    init_growth = max(init_growth, MIN_INITIAL_GROWTH_FLOOR) # Apply floor
    init_growth = min(init_growth, MAX_INITIAL_GROWTH_CEILING) # Apply ceiling
    if abs(init_growth - unbounded_growth) > 1e-6: # Check if bounding was applied
        growth_src += f" (Bounded to {format_percent(MIN_INITIAL_GROWTH_FLOOR)}-{format_percent(MAX_INITIAL_GROWTH_CEILING)})"
        projection_log.append(f"  Applied Bounds: Final Initial Growth = {format_percent(init_growth)}")
    else:
        projection_log.append(f"  Final Initial Growth Rate = {format_percent(init_growth)}")

    projection_results['initial_growth_rate'] = init_growth
    projection_results['growth_source'] = growth_src


    # --- PHASE 3: Terminal Growth Rate & FCFF Projection ---
    projection_log.append("\n--- Phase 3: Determining Terminal Growth & Projecting FCFF ---")

    # Determine Terminal Growth Rate (g)
    # Caps: Config setting, WACC fraction, final initial growth, Rf fraction
    g_init_final = init_growth if pd.notna(init_growth) and math.isfinite(init_growth) else TERMINAL_GROWTH_RATE_CAP # Use bounded initial growth
    g_wacc_cap = disc_rate * 0.60 # Cap at 60% of WACC
    g_rf_cap = final_user_rf * 0.90 # Cap at 90% of Rf
    g_config_cap = TERMINAL_GROWTH_RATE_CAP # Hard cap from config

    projection_log.append(f"\nDetermining Terminal Growth (g)...")
    projection_log.append(f"  Potential Caps: Config Limit={format_percent(g_config_cap)}, WACC*0.6={format_percent(g_wacc_cap)}, Final Initial Growth={format_percent(g_init_final)}, Rf*0.9={format_percent(g_rf_cap)}")

    # Find the minimum of all valid caps
    valid_caps = [cap for cap in [g_config_cap, g_wacc_cap, g_init_final, g_rf_cap] if np.isfinite(cap)]
    if not valid_caps:
        term_g = 0.0 # Fallback to 0 if no valid caps (shouldn't happen with config cap)
        projection_log.append("  ⚠️ No valid growth caps found. Using 0% terminal growth.")
    else:
        term_g = min(valid_caps)
        projection_log.append(f"  ↳ Minimum Cap Applied: {format_percent(term_g)}")

    # Apply floor (minimum growth rate, e.g., 0%)
    original_term_g = term_g
    term_g = max(MIN_GROWTH, term_g)
    if abs(term_g - original_term_g) > 1e-6:
        projection_log.append(f"  ↳ Floor Applied ({format_percent(MIN_GROWTH)}): Terminal Growth = {format_percent(term_g)}")

    # Ensure Terminal Growth < Discount Rate (WACC)
    if term_g >= disc_rate:
        original_term_g = term_g
        term_g = max(disc_rate - 0.005, MIN_GROWTH) # Set slightly below WACC, but not below floor
        projection_log.append(f"  ⚠️ Terminal Growth ({format_percent(original_term_g)}) was >= WACC ({format_percent(disc_rate)}). Reduced to {format_percent(term_g)}.")

    projection_log.append(f"  ▶️ Final Terminal Growth Rate (g): {format_percent(term_g)}")
    projection_results['terminal_growth_rate'] = term_g

    # Project FCFF for forecast_yrs (e.g., 3 years)
    projected_fcffs = []
    growth_rates_used = []
    pv_fcff_list = []
    cumulative_pv = 0.0
    last_fcff = precalculated_avg_basis # Start with the calculated basis (T=0)

    # Ensure effective growth rates are finite, default to 0 if not
    eff_init_g = init_growth if pd.notna(init_growth) and math.isfinite(init_growth) else 0.0
    eff_term_g = term_g if pd.notna(term_g) and math.isfinite(term_g) else 0.0

    projection_log.append(f"\n--- FCFF Projections ({forecast_yrs} Years) ---")
    projection_log.append(f"Basis (T=0 FCFF): {format_currency(last_fcff)}")
    header = f"{'Year':<5} {'Growth Rate':<15} {'Projected FCFF':<25} {'PV Factor':<15} {'PV of FCFF':<25}"
    projection_log.append(header); projection_log.append("-" * len(header))

    for yr in range(1, forecast_yrs + 1):
        # Simple linear decay from initial growth to terminal growth over the forecast period
        decay_factor = max(0, (forecast_yrs - yr)) / forecast_yrs if forecast_yrs > 0 else 0
        current_growth = eff_term_g + (eff_init_g - eff_term_g) * decay_factor
        # Ensure growth rate isn't below the minimum floor (e.g., 0%)
        current_growth = max(MIN_GROWTH, current_growth)
        growth_rates_used.append(current_growth)

        # Calculate projected FCFF
        fcff_proj = last_fcff * (1 + current_growth)
        # Handle potential numerical instability or overflow
        if not np.isfinite(fcff_proj) or abs(fcff_proj) > 1e18:
            projection_log.append(f"⚠️ Yr {yr} Projected FCFF ({fcff_proj}) invalid. Setting to 0.")
            fcff_proj = 0.0
        projected_fcffs.append(fcff_proj)
        last_fcff = fcff_proj # Update last FCFF for next year's calculation

        # Calculate Present Value Factor and PV of FCFF
        pv_factor = 1 / ((1 + disc_rate)**yr)
        pv_fcff_val = fcff_proj * pv_factor
        if not np.isfinite(pv_fcff_val): # Handle case where PV calc fails
             pv_fcff_val = 0.0
        pv_fcff_list.append(pv_fcff_val)
        cumulative_pv += pv_fcff_val

        # Log the projection for this year
        log_line = f"{yr:<5} {format_percent(current_growth):<15} {format_currency(fcff_proj):<25} {pv_factor:<15.4f} {format_currency(pv_fcff_val):<25}"
        projection_log.append(log_line)

    projection_results['projected_fcffs'] = projected_fcffs
    projection_results['growth_rates'] = growth_rates_used
    projection_results['pv_fcffs'] = pv_fcff_list
    # Ensure cumulative PV is finite
    projection_results['cumulative_pv_fcff'] = cumulative_pv if np.isfinite(cumulative_pv) else 0.0

    projection_log.append("-" * len(header))
    projection_log.append(f"Sum of Present Values of FCFFs (Years 1-{forecast_yrs}): {format_currency(projection_results['cumulative_pv_fcff'])}")


    # Calculate Terminal Value (TV)
    projection_log.append("\n--- Terminal Value Calculation ---")
    terminal_value = np.nan; pv_terminal_value = np.nan
    final_projected_fcff = projected_fcffs[-1] if projected_fcffs else np.nan # FCFF in the last forecast year (Year N)

    # Check conditions for Gordon Growth Model
    denominator = disc_rate - eff_term_g
    if denominator <= 1e-6: # WACC must be greater than terminal growth rate
        projection_results['error'] = f"WACC ({format_percent(disc_rate)}) <= Terminal Growth ({format_percent(eff_term_g)}). Cannot calculate Terminal Value."
        terminal_value = 0.0; pv_terminal_value = 0.0;
        projection_log.append(f"❌ ERROR: {projection_results['error']}. Setting TV and PV(TV) to 0.")
    elif not np.isfinite(final_projected_fcff): # Need the last projected FCFF
        projection_results['error'] = "Final projected FCFF is invalid. Cannot calculate Terminal Value."
        terminal_value = 0.0; pv_terminal_value = 0.0;
        projection_log.append(f"❌ ERROR: {projection_results['error']}. Setting TV and PV(TV) to 0.")
    else:
         # Calculate FCFF in Year N+1
         fcff_n_plus_1 = final_projected_fcff * (1 + eff_term_g)
         projection_log.append(f"FCFF in Year {forecast_yrs} (Final Forecast Year): {format_currency(final_projected_fcff)}")
         projection_log.append(f"Projected FCFF in Year {forecast_yrs + 1} (for TV Calc): {format_currency(fcff_n_plus_1)}")

         # Calculate Terminal Value at Year N
         terminal_value = fcff_n_plus_1 / denominator
         projection_log.append(f"Terminal Value (TV) at end of Year {forecast_yrs} = FCFF_N+1 / (WACC - g)")
         projection_log.append(f"  = {format_currency(fcff_n_plus_1)} / ({format_percent(disc_rate)} - {format_percent(eff_term_g)})")

         # Check if TV calculation resulted in non-finite or negative value
         if not np.isfinite(terminal_value) or terminal_value < 0:
             projection_log.append(f"⚠️ Warning: Calculated Terminal Value is invalid or negative ({format_currency(terminal_value)}). Setting TV and PV(TV) to 0.")
             if projection_results['error'] is None: projection_results['error'] = "Calculated Terminal Value invalid/negative."
             terminal_value = 0.0; pv_terminal_value = 0.0;
         else:
             projection_log.append(f"  Calculated Terminal Value (Year {forecast_yrs}): {format_currency(terminal_value)}")
             # Calculate Present Value of Terminal Value
             pv_factor_tv = 1 / ((1 + disc_rate)**forecast_yrs)
             pv_terminal_value = terminal_value * pv_factor_tv
             if not np.isfinite(pv_terminal_value): # Handle PV calc failure
                 pv_terminal_value = 0.0
                 projection_log.append(f"⚠️ Warning: PV(TV) calculation failed. Setting to 0.")
                 if projection_results['error'] is None: projection_results['error']="PV(TV) calculation failed."

             projection_log.append(f"Present Value Factor for TV (Year {forecast_yrs}): {pv_factor_tv:.4f}")
             projection_log.append(f"Present Value of Terminal Value (PV(TV)): {format_currency(pv_terminal_value)}")

    # Store results
    projection_results['terminal_value'] = terminal_value if np.isfinite(terminal_value) else 0.0
    projection_results['pv_terminal_value'] = pv_terminal_value if np.isfinite(pv_terminal_value) else 0.0

    # Display the log
    display(Markdown("```\n"+"\n".join(projection_log)+"\n```"))

    return projection_results

# --- Execute FCFF calculation and projection ---
fcff_results = {'error': 'FCFF Calculation not run.', 'initial_growth_rate': np.nan, 'growth_source':'N/A', 'terminal_growth_rate': np.nan, 'fcff_basis_method':'N/A'}

# Check if a valid method was selected and WACC is valid
if selected_fcff_method is not None and np.isfinite(discount_rate):
    # Get the pre-calculated average for the selected method
    avg_basis_chosen = precalculated_averages.get(selected_fcff_method, np.nan)

    # Ensure the basis is positive and finite
    if np.isfinite(avg_basis_chosen) and avg_basis_chosen > 1e-9:
        try:
            # Call the V21.6 function
            fcff_results = calculate_and_project_fcff_user_selected_method_v21_6(
                user_selected_method_idx=selected_fcff_method,
                precalculated_avg_basis=avg_basis_chosen,
                financials_df=stock_data.get('financials'),
                balance_sheet_df=stock_data.get('balance_sheet'),
                cashflow_df=stock_data.get('cashflow'),
                wacc_comps=wacc_components,
                disc_rate=discount_rate,
                forecast_yrs=FORECAST_YEARS, # Use the global setting (3)
                final_user_rf=risk_free_rate_input,
                default_init_growth=DEFAULT_GROWTH_RATE,
                currency_sym=currency_symbol # Pass '₹'
            )
        except Exception as e:
             display(Markdown(f"❌ **CRITICAL ERROR during FCFF Projection function call: {e}**"))
             print(traceback.format_exc())
             fcff_results={'error': f'Runtime error during projection: {e}'}
    else:
        # This case should have been caught during method selection, but double-check
        display(Markdown(f"❌ **Skipping FCFF Projection:** Basis for selected method {selected_fcff_method} is invalid or non-positive ({format_currency(avg_basis_chosen)})."))
        fcff_results={'error': f'Invalid/non-positive basis for method {selected_fcff_method}. Cannot proceed.'}
elif not np.isfinite(discount_rate):
    display(Markdown("❌ **Skipping FCFF Projection:** Invalid WACC / Discount Rate."))
    fcff_results={'error': 'Invalid WACC prevents projection.'}
elif selected_fcff_method is None:
    display(Markdown("❌ **Skipping FCFF Projection:** No valid FCFF method was selected (likely no positive average found)."))
    fcff_results={'error': 'No valid FCFF method selected. Cannot proceed.'}


# %% [markdown]
# ## 8. Enterprise Value and Equity Value Calculation
# (Section Unchanged - Uses updated helpers)

# %%
# Valuation Function Definition
def calculate_final_valuation(fcff_data, # The dict returned by the projection function
                              balance_sheet_df, # For Cash and MI
                              shares_outstanding_val, # Fetched/Validated shares
                              currency_sym, # '₹'
                              wacc_data): # WACC components dict (for debt value used)

    display(Markdown("### Final Valuation: Enterprise Value to Equity Value Bridge"))
    valuation = {
        'enterprise_value': np.nan,
        'equity_value': np.nan,
        'intrinsic_value_per_share': np.nan,
        'debt_used': np.nan,
        'cash_used': np.nan,
        'minority_interest_used': np.nan,
        'error': None
    }
    val_log = []

    # --- Check Input Validity ---
    if not isinstance(fcff_data, dict):
        valuation['error'] = "FCFF Projection Data is invalid (not a dictionary)."
        val_log.append(f"❌ FATAL ERROR: {valuation['error']}")
        display(Markdown("```\n"+"\n".join(val_log)+"\n```")); return valuation

    # Check for fatal errors reported by the FCFF function itself
    fcff_error = fcff_data.get('error')
    # Define fatal conditions from FCFF step
    is_fcff_fatal = fcff_error and ("Cannot proceed" in fcff_error or \
                                    "invalid basis" in fcff_error or \
                                    "No valid FCFF method" in fcff_error or \
                                    "WACC <= Terminal Growth" in fcff_error or \
                                    "Invalid WACC provided" in fcff_error)

    if is_fcff_fatal:
        valuation['error'] = f"Valuation halted due to fatal error from FCFF step: {fcff_error}"
        val_log.append(f"❌ FATAL ERROR: {valuation['error']}")
        display(Markdown("```\n"+"\n".join(val_log)+"\n```")); return valuation
    elif fcff_error:
        # Log non-fatal warnings from FCFF (e.g., TV calc issues but set to 0)
        val_log.append(f"ℹ️ Note: Proceeding despite non-fatal warning from FCFF step: {fcff_error}")

    # --- Calculate Enterprise Value (EV) ---
    # EV = Sum of PV(Projected FCFFs) + PV(Terminal Value)
    pv_fcff = fcff_data.get('cumulative_pv_fcff', np.nan)
    pv_tv = fcff_data.get('pv_terminal_value', np.nan)

    # Handle NaN values, defaulting to 0 if calculation failed but wasn't fatal
    if not np.isfinite(pv_fcff):
        pv_fcff = 0.0
        val_log.append(f"⚠️ PV of Forecast FCFFs invalid or N/A, assuming 0.")
    if not np.isfinite(pv_tv):
        pv_tv = 0.0
        val_log.append(f"⚠️ PV of Terminal Value invalid or N/A, assuming 0.")

    enterprise_value = pv_fcff + pv_tv
    valuation['enterprise_value'] = enterprise_value
    val_log.append(f"Sum of PV(Forecast FCFFs) (Yrs 1-{FORECAST_YEARS}): {format_currency(pv_fcff)}")
    val_log.append(f"Present Value of Terminal Value (PV(TV)): {format_currency(pv_tv)}")
    val_log.append(f"↳ Calculated Enterprise Value (EV): {format_currency(enterprise_value)}")


    # --- Bridge EV to Equity Value ---
    # Equity Value = EV - Total Debt + Cash & Cash Equivalents - Minority Interest
    debt = 0.0; cash = 0.0; minority_interest = 0.0
    debt_src = 'N/A'; cash_src = 'N/A'; mi_src = 'N/A'

    # Get Debt used in WACC calculation
    if isinstance(wacc_data, dict):
        debt = wacc_data.get('debt_value_used', 0.0)
        debt_src = wacc_data.get('debt_calc_method', '?')
        if not np.isfinite(debt):
             val_log.append(f"⚠️ Debt value from WACC components invalid ({debt}). Assuming 0.")
             debt = 0.0
             debt_src += " (Invalid->0)"
        valuation['debt_used'] = debt
    else:
        val_log.append(f"⚠️ WACC components data invalid. Assuming 0 Debt for bridge.")
        debt_src = "WACC Data Missing"
        valuation['debt_used'] = 0.0


    # Get Cash and Minority Interest from latest Balance Sheet
    if balance_sheet_df is not None and not balance_sheet_df.empty:
        bs_latest = balance_sheet_df.iloc[:, 0]
        val_log.append(f"\nBridging EV to Equity Value:")

        # Get Cash
        cash_keys = ["Cash And Cash Equivalents", "Cash", "Cash Cash Equivalents And Short Term Investments"]
        cash = safe_get(bs_latest, cash_keys, "BS(Cash)", 0)
        if not np.isfinite(cash):
            val_log.append(f"⚠️ Cash value from BS invalid ({cash}). Assuming 0.")
            cash = 0.0
            cash_src = "BS Invalid->0"
        else: cash_src = "BS Latest"
        valuation['cash_used'] = cash

        # Get Minority Interest (often zero or not present)
        mi_keys = ["Minority Interest", "Noncontrolling Interest In Consolidated Entity"]
        minority_interest = safe_get(bs_latest, mi_keys, "BS(MI)", 0)
        if not np.isfinite(minority_interest):
            # Log only if it seems invalid, otherwise MI=0 is common
            # val_log.append(f"⚠️ Minority Interest value from BS invalid ({minority_interest}). Assuming 0.")
            minority_interest = 0.0
            mi_src = "BS Invalid->0"
        else: mi_src = "BS Latest"
        valuation['minority_interest_used'] = minority_interest

        val_log.append(f"  - Total Debt ({debt_src}): {format_currency(debt)}")
        val_log.append(f"  + Cash & Equivalents ({cash_src}): {format_currency(cash)}")
        # Only show MI if it's non-zero
        if abs(minority_interest) > 1e-6:
             val_log.append(f"  - Minority Interest ({mi_src}): {format_currency(minority_interest)}")
        else:
             val_log.append(f"  - Minority Interest ({mi_src}): Approx Zero")

    else: # Balance sheet unavailable
        val_log.append("⚠️ Balance Sheet data missing. Cannot fetch Cash or Minority Interest for bridge.")
        val_log.append(f"\nBridging EV to Equity Value (using assumptions):")
        val_log.append(f"  - Total Debt ({debt_src}): {format_currency(debt)}")
        val_log.append(f"  + Cash & Equivalents (Assumed): {format_currency(cash)}")
        val_log.append(f"  - Minority Interest (Assumed): {format_currency(minority_interest)}")
        cash_src = "Assumed 0 (No BS)"
        mi_src = "Assumed 0 (No BS)"
        valuation['cash_used'] = cash
        valuation['minority_interest_used'] = minority_interest


    # Calculate Equity Value
    # Check if all components used in the bridge are finite
    if not all(np.isfinite(v) for v in [enterprise_value, debt, cash, minority_interest]):
        valuation['error'] = "One or more components for EV to Equity bridge are invalid."
        val_log.append(f"❌ ERROR: {valuation['error']}")
        valuation['equity_value'] = np.nan
        valuation['intrinsic_value_per_share'] = np.nan
        display(Markdown("```\n"+"\n".join(val_log)+"\n```")); return valuation

    equity_value = enterprise_value - debt + cash - minority_interest
    valuation['equity_value'] = equity_value
    val_log.append(f"↳ Calculated Equity Value: {format_currency(equity_value)}")

    # --- Calculate Intrinsic Value Per Share ---
    intrinsic_value_per_share = np.nan
    if not np.isfinite(equity_value):
        valuation['error'] = f"Calculated Equity Value is invalid ({format_currency(equity_value)})."
        val_log.append(f"❌ ERROR: {valuation['error']}")
    elif not np.isfinite(shares_outstanding_val) or shares_outstanding_val <= 0:
        valuation['error'] = f"Shares Outstanding is invalid ({shares_outstanding_val}). Cannot calculate IV per share."
        val_log.append(f"❌ ERROR: {valuation['error']}")
    else:
        intrinsic_value_per_share = equity_value / shares_outstanding_val
        val_log.append(f"\nCalculating Intrinsic Value per Share:")
        val_log.append(f"  Equity Value: {format_currency(equity_value)}")
        val_log.append(f"  ÷ Shares Outstanding: {shares_outstanding_val:,.0f}")
        val_log.append(f"▶️ Intrinsic Value per Share (IVPS): {format_currency(intrinsic_value_per_share)}")

    # Store final IVPS (will be NaN if calculation failed)
    valuation['intrinsic_value_per_share'] = intrinsic_value_per_share if np.isfinite(intrinsic_value_per_share) else np.nan

    # Display Log
    display(Markdown("```\n"+"\n".join(val_log)+"\n```"))
    return valuation


# --- Execute Final Valuation ---
final_valuation = {'error': 'Final Valuation calculation not run.', 'intrinsic_value_per_share': np.nan}

# Check if FCFF results and WACC components are valid dictionaries before proceeding
fcff_ready = isinstance(fcff_results, dict)
wacc_ready = isinstance(wacc_components, dict)

# Check for fatal errors from FCFF that prevent valuation
fcff_fatal_error = False
if fcff_ready:
    fcff_error_msg = fcff_results.get('error', '')
    # Define conditions that halt valuation
    if fcff_error_msg and ("Cannot proceed" in fcff_error_msg or \
                           "invalid basis" in fcff_error_msg or \
                           "No valid FCFF method" in fcff_error_msg or \
                           "WACC <= Terminal Growth" in fcff_error_msg or \
                           "Invalid WACC provided" in fcff_error_msg):
        fcff_fatal_error = True
        final_valuation['error'] = f"Valuation halted by fatal FCFF error: {fcff_error_msg}"


if fcff_ready and wacc_ready and not fcff_fatal_error:
    try:
        # Call the valuation function
        final_valuation = calculate_final_valuation(
            fcff_data=fcff_results,
            balance_sheet_df=stock_data.get('balance_sheet'),
            shares_outstanding_val=shares_outstanding, # Use the validated shares
            currency_sym=currency_symbol, # Pass '₹'
            wacc_data=wacc_components # Pass the dictionary containing debt used
        )
    except Exception as e:
        display(Markdown(f"❌ **CRITICAL ERROR during Final Valuation function call: {e}**"))
        print(traceback.format_exc())
        final_valuation = {'error': f'Runtime error during final valuation: {e}', 'intrinsic_value_per_share': np.nan}
elif fcff_fatal_error:
    # Error message already set above
    display(Markdown(f"❌ **Skipping Final Valuation:** {final_valuation['error']}"))
else:
    # If inputs themselves were invalid (not dicts)
    missing_inputs = []
    if not fcff_ready: missing_inputs.append("FCFF Results")
    if not wacc_ready: missing_inputs.append("WACC Components")
    err_msg = f"Invalid input data ({', '.join(missing_inputs)})."
    display(Markdown(f"❌ **Skipping Final Valuation:** {err_msg}"))
    final_valuation = {'error': f'Upstream data error prevents valuation: {err_msg}', 'intrinsic_value_per_share': np.nan}


# %% [markdown]
# ## 9. Financial Ratio Analysis (Ratio Plot Fix V21.6)
# (Section Unchanged - Uses updated helpers)

# %%
# Ratio Function Definition (V21.6 Plot Fixes)
# [Function analyze_ratios remains the same as provided in the previous correct version]
def analyze_ratios(stock_info_dict, financials_df, balance_sheet_df, cashflow_df,
                   enterprise_val, # Calculated EV from DCF (can be NaN)
                   current_price_val, # Fetched/validated current price (can be NaN)
                   shares_outstanding_val, # Fetched/validated shares (should be > 0)
                   currency_sym, # '₹'
                   user_benchmark_dict, # Dict of benchmark values {Ratio: Value}
                   benchmark_source_dict, # Dict of benchmark sources {Ratio: Source}
                   wacc_data): # WACC components (for debt value)

    display(Markdown("### Financial Ratio Analysis..."))
    ratios = {} # Dictionary to store calculated company ratios {Ratio: Value}
    ratio_log = [] # Log for calculation steps and warnings
    ratio_data_for_table = [] # List of lists for HTML table [Name, CompVal, BenchVal, Status]
    # Data structure for Plotly bullet chart
    ratio_plotly_data = {'Ratio': [], 'Company Value': [], 'Benchmark Value': [], 'Benchmark Source': [], 'Comparison Status': []}
    company_nm = stock_info_dict.get("shortName", ticker_symbol) # Use short name for brevity

    # --- Input Sanity Checks ---
    valid_price = current_price_val is not None and np.isfinite(current_price_val) and current_price_val > 0
    valid_shares = shares_outstanding_val is not None and np.isfinite(shares_outstanding_val) and shares_outstanding_val > 0
    valid_ev = enterprise_val is not None and np.isfinite(enterprise_val) # EV can be negative

    if not valid_price: ratio_log.append("⚠️ Current Price is invalid or zero. Price-based ratios (P/E, P/B) cannot be calculated.")
    if not valid_shares: ratio_log.append("⚠️ Shares Outstanding is invalid or zero. Per-share ratios (EPS, BVPS) and related multiples cannot be calculated.")
    if not valid_ev: ratio_log.append("⚠️ Enterprise Value (from DCF) is invalid. EV/EBITDA cannot be calculated.")

    # --- Check for Essential Financial Statements ---
    # This check is internal to the function to decide if it can proceed
    essential_missing_internal = False # Use a different name to avoid confusion
    if financials_df is None or financials_df.empty:
        ratio_log.append("❌ ERROR (Internal): Income Statement data missing. Most ratios cannot be calculated.")
        essential_missing_internal = True
    if balance_sheet_df is None or balance_sheet_df.empty:
        ratio_log.append("❌ ERROR (Internal): Balance Sheet data missing. Many ratios cannot be calculated.")
        essential_missing_internal = True
    if cashflow_df is None or cashflow_df.empty:
        # Cash flow statement is needed for Depreciation (primary source)
        ratio_log.append("⚠️ Warning (Internal): Cash Flow statement missing. Will attempt to use Depreciation from Income Statement if available.")
        # Note: We don't mark this as essential_missing=True yet, as IS fallback exists

    if essential_missing_internal:
        display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
        display(Markdown("*Cannot proceed with ratio analysis due to missing essential financial statements.*"))
        return {}, [] # Return empty results

    # --- Extract Latest Financial Data ---
    # (Rest of the analyze_ratios function is the same as before...)
    try:
        latest_fin = financials_df.iloc[:, 0]
        latest_bs = balance_sheet_df.iloc[:, 0]
        # Handle potentially missing Cash Flow DF
        latest_cf = cashflow_df.iloc[:, 0] if cashflow_df is not None and not cashflow_df.empty else pd.Series(dtype=float)
    except IndexError:
        ratio_log.append("❌ ERROR: Could not access latest column in financial statements (IndexError).")
        display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))
        return {}, []

    # --- Calculate Ratio Components ---
    ratio_log.append("--- Calculating Ratio Components from Latest Statements ---")
    # From Income Statement
    net_income = safe_get(latest_fin, ["Net Income", "Net Income Common Stockholders"], "IS(Net Income)", default=np.nan)
    revenue = safe_get(latest_fin, ["Total Revenue", "Revenue"], "IS(Revenue)", default=np.nan)
    ebit = safe_get(latest_fin, ["EBIT", "Ebit", "Operating Income"], "IS(EBIT)", default=np.nan)
    ratio_log.append(f"  Net Income: {format_currency(net_income)}, Revenue: {format_currency(revenue)}, EBIT: {format_currency(ebit)}")

    # From Cash Flow / Income Statement (Depreciation)
    dep_keys = ["Depreciation And Amortization", "Depreciation"]
    dep = safe_get(latest_cf, dep_keys, "CF(Depr)", default=np.nan)
    dep_source = "CF"
    if not np.isfinite(dep):
        ratio_log.append(f"  Depreciation not found in CF. Trying IS fallback...")
        dep_is = safe_get(latest_fin, dep_keys, "IS(Depr Fallback)", default=np.nan)
        if np.isfinite(dep_is):
            dep = dep_is
            dep_source = "IS Fallback"
            ratio_log.append(f"  Found Depreciation in IS: {format_currency(dep)}")
        else:
            dep = 0.0 # Assume zero if not found anywhere
            dep_source = "Assumed 0"
            ratio_log.append(f"  ⚠️ Depreciation not found in CF or IS. Assuming 0.")
    else:
        dep_source = "CF"
        ratio_log.append(f"  Depreciation (from {dep_source}): {format_currency(dep)}")
    dep = float(dep) # Ensure it's float

    # Calculate EBITDA
    ebitda = np.nan
    if np.isfinite(ebit) and np.isfinite(dep):
        ebitda = ebit + dep
        ratio_log.append(f"  EBITDA (EBIT + Depr): {format_currency(ebitda)}")
    elif np.isfinite(ebit):
        ebitda = ebit # If no depreciation, EBITDA = EBIT
        ratio_log.append(f"  ℹ️ EBITDA calculated as EBIT (Depreciation was {dep_source}): {format_currency(ebitda)}")
    else:
        ratio_log.append(f"  ⚠️ Cannot calculate EBITDA (EBIT is N/A). EV/EBITDA will be N/A.")

    # From Balance Sheet
    equity = safe_get(latest_bs, ["Total Stockholder Equity", "Common Stock Equity", "Stockholders Equity"], "BS(Equity)", default=np.nan)
    assets = safe_get(latest_bs, ["Total Assets"], "BS(Assets)", default=np.nan)
    curr_a = safe_get(latest_bs, ['Current Assets', 'Total Current Assets'], "BS(Current Assets)", default=np.nan)
    curr_l = safe_get(latest_bs, ['Current Liabilities', 'Total Current Liabilities'], "BS(Current Liab)", default=np.nan)
    inv = safe_get(latest_bs, ["Inventory"], "BS(Inventory)", default=np.nan) # Default to NaN if missing
    ratio_log.append(f"  Equity: {format_currency(equity)}, Assets: {format_currency(assets)}")
    ratio_log.append(f"  Current Assets: {format_currency(curr_a)}, Current Liab: {format_currency(curr_l)}")

    # Handle Inventory (assume 0 if missing/NaN for Quick Ratio)
    if not np.isfinite(inv):
        inv = 0.0
        ratio_log.append(f"  ℹ️ Inventory not found or invalid on BS. Assuming 0 for Quick Ratio calculation.")
    else:
        ratio_log.append(f"  Inventory: {format_currency(inv)}")
    inv = float(inv)

    # Get Debt (using value from WACC calculation for consistency)
    debt = 0.0; debt_src = 'N/A';
    if isinstance(wacc_data, dict):
        debt = wacc_data.get('debt_value_used', 0.0)
        debt_src = wacc_data.get('debt_calc_method', '?')
        if not np.isfinite(debt): debt = 0.0
        debt_src = f"WACC Calc ({debt_src})"
    else:
        debt = 0.0
        debt_src = "WACC Data Missing (Assumed 0)"
        ratio_log.append(f"⚠️ Using Assumed 0 Debt ({debt_src}).")
    ratio_log.append(f"  Total Debt (from {debt_src}): {format_currency(debt)}")

    # Calculate Per-Share values
    eps = (net_income / shares_outstanding_val) if valid_shares and np.isfinite(net_income) and shares_outstanding_val > 0 else np.nan
    bvps = (equity / shares_outstanding_val) if valid_shares and np.isfinite(equity) and shares_outstanding_val > 0 else np.nan
    ratio_log.append(f"  EPS: {format_currency(eps) if np.isfinite(eps) else 'N/A'}, BVPS: {format_currency(bvps) if np.isfinite(bvps) else 'N/A'}")


    # Calculate Capital Employed for ROCE
    # Method 1: Total Assets - Current Liabilities
    # Method 2: Total Equity + Total Debt
    cap_emp = np.nan; ce_src = "N/A"
    if np.isfinite(assets) and np.isfinite(curr_l):
        cap_emp_m1 = assets - curr_l
        if np.isfinite(cap_emp_m1) and cap_emp_m1 > 1.0: # Check if calculation is valid and positive
            cap_emp = cap_emp_m1
            ce_src = "Assets - Current Liab"
            ratio_log.append(f"  Capital Employed ({ce_src}): {format_currency(cap_emp)}")

    # Fallback to Method 2 if Method 1 failed or resulted non-positive
    if not np.isfinite(cap_emp):
        ratio_log.append(f"  Capital Employed (Assets - CL) failed or non-positive. Trying Equity + Debt...")
        if np.isfinite(equity) and np.isfinite(debt):
            cap_emp_m2 = equity + debt
            if np.isfinite(cap_emp_m2) and cap_emp_m2 > 1.0: # Check if valid and positive
                cap_emp = cap_emp_m2
                ce_src = "Equity + Debt"
                ratio_log.append(f"  Capital Employed ({ce_src}): {format_currency(cap_emp)}")
            else:
                 ratio_log.append(f"  Capital Employed (Equity + Debt) also failed or non-positive.")
        else:
            ratio_log.append(f"  Cannot calculate Capital Employed (Equity + Debt) due to missing components.")

    if not np.isfinite(cap_emp):
        ce_src = "Calculation Failed"
        ratio_log.append("⚠️ Capital Employed calculation failed. ROCE will be N/A.")


    # --- Calculate Ratios ---
    ratio_log.append("\n--- Calculating Final Ratios ---")
    # Helper for safe division
    def safe_div(numerator, denominator):
        num = float(numerator) if np.isfinite(numerator) else np.nan
        den = float(denominator) if np.isfinite(denominator) else np.nan
        if not np.isfinite(num) or not np.isfinite(den) or abs(den) < 1e-9: # Avoid division by zero or near-zero
            return np.nan
        try:
            return num / den
        except: # Catch any unexpected errors during division
            return np.nan

    # Valuation Ratios
    ratios['P/E'] = safe_div(current_price_val, eps) if valid_price else np.nan
    ratios['P/B'] = safe_div(current_price_val, bvps) if valid_price else np.nan
    ratios['EV/EBITDA'] = safe_div(enterprise_val, ebitda) if valid_ev and np.isfinite(ebitda) else np.nan # Check EV and EBITDA validity

    # Profitability Ratios
    ratios['ROE'] = safe_div(net_income, equity)
    ratios['ROA'] = safe_div(net_income, assets)
    ratios['ROCE'] = safe_div(ebit, cap_emp) # Uses calculated capital employed

    # Efficiency Ratio
    ratios['Asset Turnover'] = safe_div(revenue, assets)

    # Liquidity Ratios
    ratios['Current Ratio'] = safe_div(curr_a, curr_l)
    # Quick Ratio = (Current Assets - Inventory) / Current Liabilities
    quick_assets = (curr_a - inv) if np.isfinite(curr_a) and np.isfinite(inv) else np.nan
    ratios['Quick Ratio'] = safe_div(quick_assets, curr_l)

    # Log calculated ratios
    for key, val in ratios.items():
         is_perc = key in ['ROE','ROA','ROCE']
         fmt_val = format_percent(val, 1) if is_perc else (f"{val:.2f}" if np.isfinite(val) else "N/A")
         ratio_log.append(f"  {key}: {fmt_val}")


    # --- Compare Ratios to Benchmarks & Prepare Table/Plot Data ---
    display(Markdown("*(Comparing calculated ratios against benchmarks...)*"))
    for key in REQUIRED_BENCHMARK_RATIOS: # Iterate through the standard list
        comp_val = ratios.get(key) # Get the calculated value for the company
        is_perc = key in ['ROE','ROA','ROCE'] # Check if it's usually a percentage

        # Format company value for display
        val_fmt = "N/A"
        if pd.notna(comp_val) and math.isfinite(comp_val):
             val_fmt = format_percent(comp_val, 1) if is_perc else f"{comp_val:.2f}"
        elif pd.notna(comp_val): # Handle infinities
             val_fmt = "Infinity" if comp_val > 0 else "-Infinity"

        # Get benchmark value and its source
        bench_val = user_benchmark_dict.get(key, np.nan) # Benchmarks already stored as rates
        b_src = benchmark_source_dict.get(key, "?")

        # Format benchmark value for display
        bench_fmt = "N/A"
        if np.isfinite(bench_val):
             bench_fmt = format_percent(bench_val, 1) if is_perc else f"{bench_val:.2f}"
             bench_fmt += f" <small><i>({b_src.replace(' Input','').replace('Fallback ','')})</i></small>" # Add source info concisely
        else:
             bench_fmt = "<i style='color:grey'>N/A</i>"

        # Determine comparison status
        status = "<i style='color:grey'>N/A</i>"; status_color = "grey"
        plot_comp_val = np.nan; plot_bench_val = np.nan # Values for plotting (handle percentages)

        if np.isfinite(comp_val) and np.isfinite(bench_val):
            # Use values directly for non-percentages, multiply by 100 for percentages for comparison logic
            num_val = comp_val * 100 if is_perc else comp_val
            num_bench = bench_val * 100 if is_perc else bench_val
            plot_comp_val = num_val # Store value used for plotting/logic
            plot_bench_val = num_bench

            # Define colors for status (using slightly lighter shades)
            good_color = '#90EE90' # LightGreen
            bad_color = '#F08080' # LightCoral
            neutral_color = '#FFD700' # Gold
            avg_color = '#D3D3D3' # LightGray

            # Calculate relative difference, handle zero benchmark
            if abs(num_bench) > 1e-9:
                ratio_diff = (num_val - num_bench) / abs(num_bench)
            else: # Avoid division by zero if benchmark is zero
                ratio_diff = float('inf') if num_val > 0 else float('-inf') if num_val < 0 else 0.0

            # --- Status Logic (Higher is Better Ratios) ---
            if key in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio']:
                 if ratio_diff > 0.20: # Significantly better than benchmark
                     status_color = good_color; status_text = "⬆️ Stronger"
                 elif ratio_diff < -0.20: # Significantly worse than benchmark
                     status_color = bad_color; status_text = "⬇️ Weaker"
                 else: # Close to benchmark
                     status_color = avg_color; status_text = "↔️ Average"

            # --- Status Logic (Lower is Better Ratios) ---
            elif key in ['P/E','P/B','EV/EBITDA']:
                 if ratio_diff < -0.20: # Significantly lower (better) than benchmark
                     status_color = good_color; status_text = "✅ Lower (Favorable)"
                 elif ratio_diff > 0.50: # Much higher (potentially unfavorable)
                     status_color = bad_color; status_text = "‼️ Higher (Unfavorable)"
                 elif ratio_diff > 0.20: # Moderately higher (caution)
                     status_color = neutral_color; status_text = "⚠️ Elevated"
                 else: # Close to benchmark
                     status_color = avg_color; status_text = "↔️ Average"
            else: # Default for any other ratio type
                 status_text = "~ N/A ~"

            status = f"<span style='color:{status_color}; font-weight:bold;'>{status_text}</span>"

            # Add data for Plotly chart
            ratio_plotly_data['Ratio'].append(key)
            ratio_plotly_data['Company Value'].append(plot_comp_val)
            ratio_plotly_data['Benchmark Value'].append(plot_bench_val)
            ratio_plotly_data['Benchmark Source'].append(b_src)
            ratio_plotly_data['Comparison Status'].append(status_text) # Store text status for potential use

        elif not np.isfinite(comp_val):
            status = "<i style='color:grey'>N/A (Calc Failed)</i>"
            # Add placeholder data for plotting consistency if benchmark exists
            ratio_plotly_data['Ratio'].append(key)
            ratio_plotly_data['Company Value'].append(np.nan) # Company value is NaN
            ratio_plotly_data['Benchmark Value'].append(bench_val * 100 if is_perc and np.isfinite(bench_val) else bench_val)
            ratio_plotly_data['Benchmark Source'].append(b_src)
            ratio_plotly_data['Comparison Status'].append("N/A (Calc Failed)")
        elif not np.isfinite(bench_val):
            status = "<i style='color:grey'>N/A (No Benchmark)</i>"
            # Add placeholder data if company value exists but benchmark doesn't
            ratio_plotly_data['Ratio'].append(key)
            ratio_plotly_data['Company Value'].append(comp_val * 100 if is_perc else comp_val)
            ratio_plotly_data['Benchmark Value'].append(np.nan) # Benchmark value is NaN
            ratio_plotly_data['Benchmark Source'].append(b_src)
            ratio_plotly_data['Comparison Status'].append("N/A (No Benchmark)")

        # Add row to HTML table data
        ratio_data_for_table.append([f"<b>{key}</b>", val_fmt, bench_fmt, status])


    # --- Display Results ---
    # Display calculation log first
    if ratio_log: display(Markdown("```\n" + "\n".join(ratio_log) + "\n```"))

    # Display HTML Table
    if ratio_data_for_table:
        header = ["Ratio", f"{company_nm} Value", "Benchmark", "Status vs Benchmark"]
        # Basic HTML table styling
        html = "<p style='font-weight:bold;'>Ratio Comparison Summary:</p>"
        html += "<table border='1' style='width:98%; font-size:0.9em; border-collapse: collapse;'>"
        # Header row
        html += "<thead style='background-color:#e9ecef;'><tr>" + "".join(f"<th style='padding: 5px; text-align: left;'>{h}</th>" for h in header) + "</tr></thead>"
        # Body rows
        html += "<tbody>"
        for i, row in enumerate(ratio_data_for_table):
            bg_color = "#f8f9fa" if i % 2 == 0 else "#ffffff" # Alternating row colors
            html += f"<tr style='background-color:{bg_color};'>"
            # Cell formatting
            html += f"<td style='padding: 5px;'>{row[0]}</td>" # Ratio Name
            html += f"<td style='padding: 5px; text-align: right;'>{row[1]}</td>" # Company Value
            html += f"<td style='padding: 5px; text-align: right;'>{row[2]}</td>" # Benchmark Value
            html += f"<td style='padding: 5px; text-align: center;'>{row[3]}</td>" # Status
            html += "</tr>"
        html += "</tbody></table><br>"
        display(HTML(html))
    else:
        display(Markdown("*No ratio data available to display in table.*"))


    # --- Ratio Comparison Chart (Plotly Bullet Chart - V21.6 Fixes) ---
    if ratio_plotly_data['Ratio'] and any(np.isfinite(v) for v in ratio_plotly_data['Company Value']) and any(np.isfinite(v) for v in ratio_plotly_data['Benchmark Value']):
       valid_plot_indices = [i for i, (cv, bv) in enumerate(zip(ratio_plotly_data['Company Value'], ratio_plotly_data['Benchmark Value'])) if np.isfinite(cv) and np.isfinite(bv)]

       if not valid_plot_indices:
            display(Markdown("*No ratios with both valid Company and Benchmark values found for plotting.*"))
       else:
            num_ratios_to_plot = len(valid_plot_indices)
            fig_ratios = make_subplots(rows=num_ratios_to_plot, cols=1,
                                       subplot_titles=[ratio_plotly_data['Ratio'][i] for i in valid_plot_indices],
                                       vertical_spacing=0.05 + 0.05/num_ratios_to_plot) # Dynamic spacing

            # Define color scheme based on whether higher or lower is better
            def get_range_and_delta_colors(ratio_key):
                # Define colors for status (using slightly lighter shades)
                good_color = '#90EE90' # LightGreen
                bad_color = '#F08080' # LightCoral
                neutral_color = '#FFD700' # Gold
                avg_color = '#D3D3D3' # LightGray
                # Higher is better: Green > Yellow > Red
                if ratio_key in ['ROE','ROA','ROCE','Asset Turnover','Current Ratio','Quick Ratio']:
                    colors = [bad_color, avg_color, good_color] # Red, Gray, Green for ranges below, around, above benchmark
                    decreasing_color = bad_color # Red if value decreases vs benchmark
                    increasing_color = good_color # Green if value increases vs benchmark
                    delta_threshold = 0.20 # 20% deviation considered significant
                # Lower is better: Green < Yellow < Red
                elif ratio_key in ['P/E','P/B','EV/EBITDA']:
                    colors = [good_color, avg_color, bad_color] # Green, Gray, Red for ranges below, around, above benchmark
                    decreasing_color = good_color # Green if value decreases vs benchmark
                    increasing_color = bad_color # Red if value increases vs benchmark
                    delta_threshold = 0.20 # 20% deviation considered significant
                else: # Default (neutral)
                    colors = ['lightgrey','darkgrey','grey']
                    decreasing_color = 'grey'
                    increasing_color = 'grey'
                    delta_threshold = 0.20
                return colors, decreasing_color, increasing_color, delta_threshold

            plot_row = 1 # Plotly subplot rows are 1-indexed
            for i in valid_plot_indices:
                name = ratio_plotly_data['Ratio'][i]
                comp_v = ratio_plotly_data['Company Value'][i]
                bench_v = ratio_plotly_data['Benchmark Value'][i]
                b_src_p = ratio_plotly_data['Benchmark Source'][i]
                is_perc = name in ['ROE','ROA','ROCE']

                range_colors, dec_color, inc_color, delta_thresh = get_range_and_delta_colors(name)

                # Determine gauge range dynamically based on values
                val_range = abs(comp_v - bench_v)
                padding = max(val_range * 0.5, abs(bench_v * 0.3), 5 if is_perc else 1) # Add padding around values, ensure min padding
                gauge_min = min(comp_v, bench_v) - padding
                gauge_max = max(comp_v, bench_v) + padding

                # Apply specific overrides for certain ratios to keep ranges reasonable
                if name in ['P/E','EV/EBITDA']: gauge_min = max(gauge_min, -20); gauge_max = min(gauge_max, 100)
                elif name == 'P/B': gauge_min = max(gauge_min, -5); gauge_max = min(gauge_max, 20)
                elif name in ['Current Ratio','Quick Ratio']: gauge_min = max(gauge_min, 0); gauge_max = min(gauge_max, 5)
                elif is_perc: gauge_min = max(gauge_min, -100); gauge_max = min(gauge_max, 150)
                elif name == 'Asset Turnover': gauge_min = max(gauge_min, -0.5); gauge_max = min(gauge_max, 3)

                # Ensure min < max
                if gauge_min >= gauge_max: gauge_max = gauge_min + max(1, abs(gauge_min * 0.1))

                # Define steps for the bullet gauge background colors
                # Ranges around the benchmark value
                step_delta = abs(bench_v * delta_thresh) if abs(bench_v)>1e-6 else (1 if is_perc else 0.1) # Size of the middle range
                bound1 = bench_v - step_delta
                bound2 = bench_v + step_delta
                steps_config = [
                    {'range': [gauge_min, bound1], 'color': range_colors[0]},
                    {'range': [bound1, bound2], 'color': range_colors[1]},
                    {'range': [bound2, gauge_max], 'color': range_colors[2]}
                ]
                # Filter out steps with non-increasing ranges
                steps_config = [s for s in steps_config if s['range'][0] < s['range'][1]]
                if not steps_config: # Fallback if ranges collapse
                    steps_config = [{'range': [gauge_min, gauge_max], 'color':'lightgrey'}]

                # Format benchmark display text
                bench_fmt_plot = (f"{bench_v:.1f}%" if is_perc else f"{bench_v:.2f}") + f" <small><i>({b_src_p.replace('Fallback ','')})</i></small>"
                # Use subplot title for name, keep indicator title simpler
                indicator_title = f"<i style='font-size:0.8em'>(Benchmark: {bench_fmt_plot})</i>"
                value_format = ".1f" if is_perc else ".2f" # Formatting for numbers in the indicator

                fig_ratios.add_trace(go.Indicator(
                    mode = "number+gauge+delta",
                    value = comp_v,
                    number = {'suffix': "%" if is_perc else "", 'valueformat': value_format},
                    delta = {'reference': bench_v, 'relative': False, 'valueformat': value_format,
                             'decreasing': {'color': dec_color}, 'increasing': {'color': inc_color}},
                    domain = {'row': plot_row - 1, 'column': 0}, # Domain uses 0-based index
                    title = {'text': indicator_title, 'font': {'size': 10}}, # Smaller font for benchmark
                    gauge = {
                        'shape': "bullet",
                        'axis': {'range': [gauge_min, gauge_max], 'tickformat': value_format},
                        'threshold': { # Line showing the benchmark
                            'line': {'color': "black", 'width': 2},
                            'thickness': 0.75,
                            'value': bench_v
                        },
                        'steps': steps_config,
                        'bar': {'color': "#2c3e50"} # Color of the main value bar
                    }
                ), row=plot_row, col=1) # Add trace uses 1-based index
                plot_row += 1 # Move to the next row for the next ratio

            fig_ratios.update_layout(
                height=max(300, num_ratios_to_plot * 85), # Adjust height based on number of ratios
                title_text='Ratio Performance vs. Benchmarks',
                title_x=0.5,
                margin=dict(l=180, r=40, t=60 + num_ratios_to_plot * 15 , b=30) # Adjust top margin for subplot titles
            )
            fig_ratios.show()
    else:
        display(Markdown("*No comparable ratio data found for generating the performance chart.*"))

    return ratios, ratio_data_for_table # Return calculated ratios and table data


# --- Execute Ratio Analysis ---
ratios_calculated = {}; ratio_table = []

# Check if prerequisites for ratio analysis are met
valuation_results_exist = isinstance(final_valuation, dict)
wacc_components_exist = isinstance(wacc_components, dict)

# *** FIX START ***
# Check for presence of essential statements BEFORE the conditional call
financials_present = stock_data.get('financials') is not None and not stock_data['financials'].empty
balance_sheet_present = stock_data.get('balance_sheet') is not None and not stock_data['balance_sheet'].empty
# This variable determines if the essential statements needed by analyze_ratios are available
required_statements_present = financials_present and balance_sheet_present
# *** FIX END ***

# Define if valuation produced a usable (even if NaN) EV or failed fatally
valuation_ok_for_ratios = valuation_results_exist and not (final_valuation.get('error') and "halted" in final_valuation.get('error',''))

# *** FIX: Use 'required_statements_present' instead of 'essential_missing' in the condition ***
if valuation_ok_for_ratios and wacc_components_exist and required_statements_present:
    try:
        # Get EV from valuation results, allow NaN if calculation failed but wasn't fatal
        ev_for_ratios = final_valuation.get('enterprise_value', np.nan)
        if not np.isfinite(ev_for_ratios):
             # This warning will be logged inside analyze_ratios as well
             print("INFO (Pre-call): Enterprise Value from DCF is invalid/NaN. EV/EBITDA ratio will be N/A.")

        # Call the ratio analysis function
        ratios_calculated, ratio_table = analyze_ratios(
            stock_info_dict=stock_data['info'],
            financials_df=stock_data.get('financials'),
            balance_sheet_df=stock_data.get('balance_sheet'),
            cashflow_df=stock_data.get('cashflow'), # Pass CF even if empty, analyze_ratios handles it
            enterprise_val=ev_for_ratios,
            current_price_val=current_price,
            shares_outstanding_val=shares_outstanding,
            currency_sym=currency_symbol, # '₹'
            user_benchmark_dict=user_benchmarks,
            benchmark_source_dict=benchmark_sources,
            wacc_data=wacc_components # Pass WACC components for debt info
        )
    except Exception as e:
        display(Markdown(f"❌ **CRITICAL ERROR during Ratio Analysis function call: {e}**"))
        print(traceback.format_exc())
        ratios_calculated = {'error': f'Runtime error during ratio analysis: {e}'}
else:
    # Determine reason for skipping
    reason = []
    # *** FIX: Update the reason check using the correct variable ***
    if not required_statements_present:
        missing_stmts = []
        if not financials_present: missing_stmts.append("Income Statement")
        if not balance_sheet_present: missing_stmts.append("Balance Sheet")
        reason.append(f"Missing essential financial statement(s): {', '.join(missing_stmts)}")
    if not valuation_ok_for_ratios: reason.append(f"Valuation failed fatally ({final_valuation.get('error', 'Unknown')})")
    if not wacc_components_exist: reason.append("WACC components calculation failed")
    error_reason = "; ".join(reason) if reason else "Unknown upstream errors" # Use semicolon for clarity
    display(Markdown(f"❌ **Skipping Ratio Analysis:** {error_reason}"))
    ratios_calculated = {'error': f'Skipped due to: {error_reason}'}


# %% [markdown]
# ## 10. Summary & Visualization Dashboard
# (Section Unchanged - Uses updated helpers)

# %%
# [Dashboard function display_summary_dashboard remains the same as previous correct version]
# Dashboard Function Definition
def display_summary_dashboard(company_nm, ticker_sym, current_pr, intrinsic_val_per_sh,
                              wacc_val, term_growth, init_growth, growth_src,
                              final_val_dict, # Full valuation dict for EV components
                              hist_df, # Price history dataframe
                              fcff_res, # Full FCFF results dict
                              currency_sym, # '₹'
                              info_dict, # Stock info dict
                              input_src_dict): # Input sources dict

    display(Markdown(f"## Valuation Summary Dashboard: {company_nm} ({ticker_sym})"))

    # --- Summarize Input Sources ---
    sources = []
    # Concise source summary
    key_map = {'Rf':'Rf', 'Rm':'Rm', 'Beta':'Beta', 'Credit Spread':'Spread', 'Benchmarks':'Benchmarks', 'FCFF Method':'FCFF Method'}
    for key, mapped_key in key_map.items():
        source_val = input_src_dict.get(key, '?')
        # Clean up source descriptions for brevity
        if isinstance(source_val, str):
            source_clean = source_val.split('(')[0].strip().replace('User Input/Default', 'User/Def').replace('User Input', 'User').replace('Dynamic Fetch', 'Dyn').replace('Default','Def').replace('Derived','Deriv')
            sources.append(f"{mapped_key}:`{source_clean}`")
    source_summary = " | ".join(sources)

    # Growth and Basis Info
    growth_disp = growth_src if isinstance(growth_src,str) else 'N/A'
    # Clean growth source
    growth_disp_clean = growth_disp.split('[')[0].strip().replace('Geometric Mean','GM').replace('valid CAGR(s)','')
    fcff_basis_disp = fcff_res.get('fcff_basis_method', 'N/A') if isinstance(fcff_res, dict) else 'N/A'
    fcff_basis_clean = fcff_basis_disp.split('(User Selected')[0].strip() if isinstance(fcff_basis_disp, str) else 'N/A'

    display(Markdown(f"**Key Inputs:** {source_summary} | **Init Growth:** `{growth_disp_clean}` | **FCFF Basis:** `{fcff_basis_clean}`"))

    # --- Check Validity of Key Metrics ---
    valid_iv = intrinsic_val_per_sh is not None and np.isfinite(intrinsic_val_per_sh)
    valid_cp = current_pr is not None and np.isfinite(current_pr) and current_pr > 0
    valid_w = wacc_val is not None and np.isfinite(wacc_val)
    valid_tg = term_growth is not None and np.isfinite(term_growth)
    valid_ig = init_growth is not None and np.isfinite(init_growth)

    # --- Create Indicator Subplots ---
    fig = make_subplots(
        rows=1, cols=3,
        specs=[[{'type': 'indicator'}, {'type': 'indicator'}, {'type': 'indicator'}]],
        subplot_titles=("Price vs. Intrinsic Value", "WACC", "Growth Rates"),
        horizontal_spacing=0.15 # Increase spacing slightly
    )

    # Indicator 1: Price vs. Intrinsic Value
    iv_num = intrinsic_val_per_sh if valid_iv else np.nan
    iv_fmt = format_currency(iv_num) if valid_iv else "N/A"
    cp_num = current_pr if valid_cp else 0.0 # Use 0 if current price invalid for display
    delta_config = None # Configuration for the delta value display
    comparison_text = f"Intrinsic Value: {iv_fmt}"
    mode = "number" # Default mode if no comparison possible

    if valid_iv and valid_cp:
        mode = "number+delta" # Add delta if comparison is possible
        if iv_num > 1e-9: # Ensure IV is positive for meaningful percentage comparison
            upside = (iv_num / cp_num) - 1
            comparison_text += f" (Upside: {upside:+.1%})"
            # Delta shows difference *from* the reference (IV) *to* the value (Current Price)
            delta_config = {'reference': iv_num, 'relative': False, 'valueformat': '.2f',
                            'prefix': currency_sym, # Show currency on delta
                            'decreasing': {'color': '#2ecc71'}, # Price below IV (good) -> Green
                            'increasing': {'color': '#e74c3c'}  # Price above IV (bad) -> Red
                           }
        else: # IV is zero or negative
            comparison_text += " (≤ 0)"
            delta_config = {'reference': 0, 'relative': False, 'valueformat': '.2f', 'prefix': currency_sym,
                            'increasing': {'color': '#e74c3c'} # Price above 0 -> Red
                           }
    elif valid_iv:
        comparison_text += " (Current Price N/A)"
    else: # IV is invalid
        comparison_text = f"Intrinsic Value: N/A (Calculation Error)"

    fig.add_trace(go.Indicator(
        mode=mode,
        value=cp_num, # Display Current Price as the main value
        number={'prefix': currency_sym, 'valueformat': '.2f'}, # Format main number
        title={"text": f"<span style='font-size:0.8em'>{comparison_text}</span>", 'font_size': 12},
        delta=delta_config, # Add the delta comparison if configured
        domain={'row': 0, 'column': 0}
    ), row=1, col=1)

    # Indicator 2: WACC
    wacc_num = wacc_val * 100 if valid_w else None # Display as percentage
    gauge_config_wacc = None
    if valid_w:
        gauge_config_wacc = {
            'axis': {'range': [MIN_WACC * 100, MAX_WACC * 100], 'tickformat': '.1f'}, # Use bounds from config
            'bar': {'color': "darkblue"},
            'steps': [ # Color steps for context (e.g., low, medium, high WACC ranges)
                {'range': [MIN_WACC * 100, 9], 'color': "#d5f5e3"}, # Light Green up to 9%
                {'range': [9, 14], 'color': "#fdebd0"},          # Light Yellow 9-14%
                {'range': [14, MAX_WACC * 100], 'color': "#fadbd8"} # Light Red above 14%
            ],
             'threshold': { # Optional line at the calculated value
                 'line': {'color': "red", 'width': 2}, 'thickness': 0.75, 'value': wacc_num}
        }

    fig.add_trace(go.Indicator(
        mode="gauge+number",
        value=wacc_num,
        number={'suffix': "%", 'valueformat': '.1f' if valid_w else ''},
        gauge=gauge_config_wacc,
        title={"text": "<span style='font-size:0.8em'>WACC</span>", 'font_size': 12},
        domain={'row': 0, 'column': 1}
    ), row=1, col=2)


    # Indicator 3: Growth Rates (Terminal Growth shown, Initial Growth in title)
    tg_num = term_growth * 100 if valid_tg else None
    ig_fmt = format_percent(init_growth, 1) if valid_ig else "N/A"
    g_title = f"<span style='font-size:0.8em'>Terminal Growth (g)<br>(Initial: {ig_fmt})</span>"
    gauge_config_growth = None
    if valid_tg:
        # Dynamic range based on terminal growth, but ensure some padding and min/max
        g_range_min = min(MIN_GROWTH * 100, tg_num - 2) # At least 2% below tg or min growth
        g_range_max = max(TERMINAL_GROWTH_RATE_CAP * 100 + 1, tg_num + 2) # At least 2% above tg or config cap+1
        gauge_config_growth = {
            'axis': {'range': [g_range_min, g_range_max], 'tickformat': '.1f'},
            'bar': {'color': "darkgrey"},
            'steps': [ # Color steps for context (e.g., low, sustainable, high terminal growth)
                {'range': [g_range_min, 2.0], 'color': "#fadbd8"}, # Light Red below 2%
                {'range': [2.0, 4.5], 'color': "#fdebd0"},      # Light Yellow 2-4.5%
                {'range': [4.5, g_range_max], 'color': "#d5f5e3"} # Light Green above 4.5%
            ],
            'threshold': {'line': {'color': "black", 'width': 2}, 'thickness': 0.75, 'value': tg_num}
        }

    fig.add_trace(go.Indicator(
        mode="gauge+number",
        value=tg_num,
        number={'suffix': "%", 'valueformat': '.1f' if valid_tg else ''},
        gauge=gauge_config_growth,
        title={"text": g_title, 'font_size': 12},
        domain={'row': 0, 'column': 2}
    ), row=1, col=3)

    fig.update_layout(
        height=250, # Increased height slightly
        margin=dict(l=30, r=30, t=60, b=20) # Adjust margins, esp. top for titles
    )
    fig.show()


    # --- Price History Chart ---
    if hist_df is not None and not hist_df.empty:
        price_col = None
        # Prefer 'Adj Close' if available and has data, otherwise 'Close'
        if 'Adj Close' in hist_df.columns and hist_df['Adj Close'].notna().any():
            price_col = 'Adj Close'
        elif 'Close' in hist_df.columns and hist_df['Close'].notna().any():
            price_col = 'Close'

        if price_col:
             fig_h = px.line(hist_df, x=hist_df.index, y=price_col,
                             title=f'{company_nm} Price History ({price_col}) vs. DCF Intrinsic Value',
                             labels={'x': 'Date', price_col: f'Price ({currency_sym})'})
             fig_h.update_layout(height=400, showlegend=False)
             fig_h.update_xaxes(rangeslider_visible=False) # Hide rangeslider for cleaner look

             # Add horizontal line for Intrinsic Value
             if valid_iv:
                 fig_h.add_hline(y=iv_num, line_dash="dash", line_color="#e74c3c", # Red dashed line
                                 annotation_text=f"DCF Value: {format_currency(iv_num)}",
                                 annotation_position="bottom right", annotation_font_color="#e74c3c")

             # Add horizontal line for Current Price
             if valid_cp:
                 fig_h.add_hline(y=cp_num, line_dash="dot", line_color="#3498db", # Blue dotted line
                                 annotation_text=f"Current Price: {format_currency(cp_num)}",
                                 annotation_position="top left", annotation_font_color="#3498db")

             fig_h.show()
        else:
             display(Markdown("*Could not find 'Adj Close' or 'Close' column in historical data for plotting.*"))
    else:
        display(Markdown("*Historical price data not available for plotting.*"))


    # --- EV Composition Pie Chart ---
    # Check if FCFF results are valid and contain the necessary components
    fcff_data_valid_for_pie = isinstance(fcff_res, dict)
    fcff_fatal_pie = fcff_data_valid_for_pie and fcff_res.get('error') and ("halted" in fcff_res.get('error', '') or "Cannot proceed" in fcff_res.get('error', '')) # Check for fatal errors specifically

    if fcff_data_valid_for_pie and not fcff_fatal_pie:
        pv_f = fcff_res.get('cumulative_pv_fcff', np.nan)
        pv_t = fcff_res.get('pv_terminal_value', np.nan)
        # Check if components are valid numbers
        valid_pvs = [v for v in [pv_f, pv_t] if np.isfinite(v)]
        total_ev = sum(valid_pvs) if valid_pvs else np.nan # Sum only the valid components

        if np.isfinite(total_ev): # Check if total EV calculation was successful
            labels=[]; values=[]; pulls=[]; colors=['#4e79a7','#f28e2b'] # Define colors

            # Add PV of Forecast FCFFs if significant
            if abs(pv_f) > 1e-6:
                labels.append(f'PV({FORECAST_YEARS}yr Forecast FCFFs)')
                values.append(abs(pv_f)) # Use absolute value for pie chart size
                pulls.append(0)
            # Add PV of Terminal Value if significant
            if abs(pv_t) > 1e-6:
                labels.append('PV(Terminal Value)')
                values.append(abs(pv_t))
                pulls.append(0.05 if len(labels) > 1 else 0) # Slightly pull TV slice if both exist

            if labels: # If there's anything to plot
                hover_template = f"<b>%{{label}}</b><br>Value Contribution:{currency_sym} %{{value:,.0f}}<br>Percentage of EV: %{{percent}}<extra></extra>"
                fig_pie = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4,
                                                 pull=pulls, marker_colors=colors[:len(labels)],
                                                 hovertemplate=hover_template,
                                                 textinfo='percent+label', sort=False, rotation=90)]) # Rotate start
                fig_pie.update_layout(title_text=f"DCF Enterprise Value Breakdown (Total EV: {format_currency(total_ev)})",
                                      title_x=0.5, height=350, showlegend=False, # Hide legend, info is on slices
                                      margin=dict(t=60, b=30, l=30, r=30))
                fig_pie.show()
            else:
                display(Markdown(f"*DCF components (PV FCFF, PV TV) are zero or negligible. Cannot generate EV breakdown chart.*"))
        elif fcff_fatal_pie: # If EV calc failed due to fatal FCFF error that halted things
             display(Markdown(f"*Cannot generate EV breakdown chart due to fatal FCFF error: {fcff_res.get('error')}*"))
        else: # If EV calc failed for other reasons (e.g., both pv_f and pv_t were NaN but not fatal)
             display(Markdown(f"*Enterprise Value calculation failed ({format_currency(total_ev)}). Cannot generate EV breakdown chart.*"))
    else: # If fcff_res itself is invalid
        display(Markdown(f"*FCFF results data is invalid. Cannot generate EV breakdown chart.*"))


    # --- Projected FCFF Bar Chart ---
    fcff_proj_data_valid = isinstance(fcff_res, dict) and not fcff_fatal_pie
    if fcff_proj_data_valid:
        proj_fcffs = fcff_res.get('projected_fcffs', [])
        pv_proj_fcffs = fcff_res.get('pv_fcffs', [])
        num_proj = len(proj_fcffs)

        if num_proj > 0 and len(pv_proj_fcffs) == num_proj:
            years = [f'Year {i+1}' for i in range(num_proj)]
            # Ensure data is numeric, replace NaN/inf with 0 for plotting
            fcff_clean = [f if np.isfinite(f) else 0 for f in proj_fcffs]
            pv_clean = [p if np.isfinite(p) else 0 for p in pv_proj_fcffs]

            df_proj = pd.DataFrame({'Year': years, 'Projected FCFF': fcff_clean, 'PV of FCFF': pv_clean})
            # Melt dataframe for easy plotting with Plotly Express
            df_melt = df_proj.melt(id_vars='Year', var_name='Cash Flow Type', value_name='Value')

            fig_bar = px.bar(df_melt, x='Year', y='Value', color='Cash Flow Type',
                             title=f'Projected FCFF & Present Values ({FORECAST_YEARS} Years)',
                             labels={'Value': f'Value ({currency_sym})', 'Year': 'Forecast Year'},
                             barmode='group', # Group bars side-by-side
                             height=350,
                             color_discrete_map={'Projected FCFF': '#5dade2', 'PV of FCFF': '#af7ac5'}) # Custom colors

            hover_template_bar = f"<b>%{{customdata[0]}}</b><br>Year: %{{x}}<br>Value: {currency_sym} %{{y:,.0f}}<extra></extra>"
            fig_bar.update_traces(hovertemplate=hover_template_bar, customdata=df_melt[['Cash Flow Type']])
            fig_bar.update_layout(yaxis_title=f"Value ({currency_sym})", legend_title="Cash Flow Type")
            fig_bar.show()
        elif num_proj == 0:
             display(Markdown("*No projected FCFF data available to plot.*"))
        else: # Data mismatch
             display(Markdown("*Mismatch between projected FCFF and PV data lengths. Cannot plot forecast.*"))
    elif fcff_fatal_pie:
        display(Markdown(f"*Cannot generate FCFF forecast chart due to fatal FCFF error.*"))
    else:
        display(Markdown(f"*FCFF projection results invalid. Cannot generate forecast chart.*"))


    # --- Valuation Conclusion Text ---
    display(Markdown("### Valuation Conclusion"));
    conclusion = ""
    if isinstance(final_val_dict, dict):
        iv = final_val_dict.get('intrinsic_value_per_share', np.nan)
        v_err = final_val_dict.get('error') # Error from the valuation step itself
        f_err = fcff_res.get('error') if isinstance(fcff_res,dict) else "FCFF Data Invalid" # Error from FCFF step

        # Prioritize fatal errors
        if v_err and "halted" in v_err: conclusion = f"<span style='color:red; font-weight:bold;'>❌ Valuation Halted: {v_err}</span>"
        elif fcff_fatal_pie: conclusion = f"<span style='color:red; font-weight:bold;'>❌ Valuation Halted (FCFF Error): {f_err}</span>"
        elif v_err: conclusion = f"<span style='color:red; font-weight:bold;'>❌ Valuation Error: {v_err}</span>" # Non-fatal val error
        elif not valid_iv: conclusion=f"<span style='color:red; font-weight:bold;'>❌ Valuation Error: Intrinsic Value calculation failed.</span>"
        # Valuation seems successful, compare IV to Price
        elif valid_iv and iv <= 0: conclusion=f"<span style='color:orange; font-weight:bold;'>⚠️ Negative/Zero DCF Value: {format_currency(iv)}.</span> Model assumptions may be flawed or company generates negative value under these assumptions."
        elif valid_iv and not valid_cp: conclusion=f"✅ DCF Intrinsic Value Calculated: **{format_currency(iv)}**. Cannot compare to current price (N/A)."
        elif valid_iv and valid_cp:
             diff = (iv / cp_num) - 1 # Upside/Downside percentage
             cp_fmt = format_currency(cp_num)
             iv_fmt = format_currency(iv)
             margin = 0.15 # Define margin for "Fairly Valued" (e.g., +/- 15%)

             if abs(diff) < margin:
                 conclusion = f"<span style='color:grey; font-weight:bold;'>↔️ Fairly Valued (within +/-{margin:.0%}):</span> Current Price ({cp_fmt}) is close to DCF Intrinsic Value ({iv_fmt}). Difference: **{diff:+.1%}**"
             elif diff < 0: # Price is higher than IV -> Overvalued
                 over_pct = (cp_num / iv) - 1 if iv != 0 else float('inf') # Premium = (Price/IV)-1
                 conclusion = f"<span style='color:#e74c3c; font-weight:bold;'>🔻 Potentially Overvalued:</span> Current Price ({cp_fmt}) is **{over_pct:+.1%}** higher than DCF Intrinsic Value ({iv_fmt})."
             else: # Price is lower than IV -> Undervalued
                 conclusion = f"<span style='color:#2ecc71; font-weight:bold;'>✅ Potentially Undervalued:</span> Current Price ({cp_fmt}) is **{diff:+.1%}** lower than DCF Intrinsic Value ({iv_fmt})."
        else: # Should not happen if logic above is correct
            conclusion=f"<span style='color:red; font-weight:bold;'>❌ Valuation Error: Unexpected state.</span>"
    else: # final_val_dict was not a dict
        conclusion=f"<span style='color:red; font-weight:bold;'>❌ Valuation Error: Final valuation data structure invalid.</span>"

    # Display the final conclusion text, wrapped in a quote block
    display(Markdown(f"> **{conclusion}**"))


# --- Execute Summary Dashboard Generation ---
# Check if valuation was attempted and didn't halt fatally
valuation_successful = (isinstance(final_valuation, dict) and not (final_valuation.get('error') and "halted" in final_valuation.get('error','')))

# Check if other necessary data is available
fcff_data_exists = isinstance(fcff_results, dict)
info_data_exists = isinstance(stock_data.get('info'), dict)
input_sources_exist = isinstance(input_sources, dict)
history_data_exists = stock_data.get('history') is not None # Check if history df exists

if valuation_successful and fcff_data_exists and info_data_exists and input_sources_exist and history_data_exists:
    try:
        # Call the dashboard function with all necessary components
        display_summary_dashboard(
            company_nm=company_name,
            ticker_sym=ticker_symbol,
            current_pr=current_price,
            intrinsic_val_per_sh=final_valuation.get('intrinsic_value_per_share'),
            wacc_val=discount_rate,
            term_growth=fcff_results.get('terminal_growth_rate'),
            init_growth=fcff_results.get('initial_growth_rate'),
            growth_src=fcff_results.get('growth_source', 'N/A'),
            final_val_dict=final_valuation, # Pass the whole dict for EV breakdown
            hist_df=stock_data['history'], # Pass price history df
            fcff_res=fcff_results, # Pass the whole dict for EV/FCFF charts
            currency_sym=currency_symbol, # '₹'
            info_dict=stock_data['info'],
            input_src_dict=input_sources
        )
    except Exception as e:
        display(Markdown(f"❌ **CRITICAL ERROR during Dashboard Generation: {e}**"))
        print(traceback.format_exc())
else:
    # Determine why dashboard is skipped
    skip_reasons = []
    if not valuation_successful: skip_reasons.append(f"Valuation failed or halted ({final_valuation.get('error', 'Unknown')})")
    if not fcff_data_exists: skip_reasons.append("FCFF results missing")
    if not info_data_exists: skip_reasons.append("Stock info missing")
    if not input_sources_exist: skip_reasons.append("Input sources tracking missing")
    if not history_data_exists: skip_reasons.append("Price history data missing")
    skip_reason_str = ", ".join(skip_reasons) if skip_reasons else "Unknown reason"
    display(Markdown(f"❌ **Skipping Summary Dashboard:** {skip_reason_str}"));


# %% [markdown]
# ## 11. Assumptions & Disclaimer (Reflects V21.6 Logic)
# (Section Unchanged - Text reflects logic)

# %%
# [Assumption generation code remains the same]
# --- Dynamically Generate Assumption Strings from Results ---
# Use .get with defaults for robustness in case dicts are missing keys

# Input sources (already fetched)
rf_src = input_sources.get('Rf','?')
rm_src = input_sources.get('Rm','?')
beta_src = input_sources.get('Beta','?')
cs_src = input_sources.get('Credit Spread','N/A')
bench_src = input_sources.get('Benchmarks','?')
fcff_method_src = input_sources.get('FCFF Method','N/A')

# From FCFF results
growth_src = fcff_results.get('growth_source','N/A') if isinstance(fcff_results,dict) else 'Error retrieving growth source'
term_g = fcff_results.get('terminal_growth_rate',np.nan) if isinstance(fcff_results,dict) else np.nan
init_g = fcff_results.get('initial_growth_rate',np.nan) if isinstance(fcff_results,dict) else np.nan
basis_meth = fcff_results.get('fcff_basis_method','N/A') if isinstance(fcff_results,dict) else 'Error retrieving FCFF basis method'

# From WACC results
tax_src = wacc_components.get('tax_rate_source','N/A') if isinstance(wacc_components,dict) else 'Error retrieving tax source'
eq_src = wacc_components.get('equity_source','N/A') if isinstance(wacc_components,dict) else 'Error retrieving equity source'
wgt_src = wacc_components.get('weights_source','N/A') if isinstance(wacc_components,dict) else 'Error retrieving weights source'
kd_method = wacc_components.get('debt_calc_method','N/A') if isinstance(wacc_components,dict) else 'Error retrieving Kd method'

# Format rates for display
term_g_fmt = format_percent(term_g) if np.isfinite(term_g) else "N/A"
init_g_fmt = format_percent(init_g) if np.isfinite(init_g) else "N/A"
def_init_g_fmt = format_percent(DEFAULT_GROWTH_RATE)
min_init_g_fmt = format_percent(MIN_INITIAL_GROWTH_FLOOR)
max_init_g_fmt = format_percent(MAX_INITIAL_GROWTH_CEILING)

# Clean up source descriptions for display
def clean_source(source_str):
    if not isinstance(source_str, str): return str(source_str)
    return source_str.split('(')[0].strip().replace('User Input/Default', 'User/Default').replace('User Input', 'User').replace('Dynamic Fetch', 'Dynamic').replace('Default','Default').replace('Derived','Derived').replace('Fallback ','')

# %%
# [Disclaimer text remains the same]
# ----> Display Assumptions & Disclaimer (Updated Text Reflects V21.6 Logic) <----
display(Markdown(f"""
**Key Assumptions & Input Sources:**

*   **Financial Data:** Sourced primarily from Yahoo Finance via `yfinance`. Data accuracy and availability are subject to the provider. Quarterly data used as fallback if annual data is missing.
*   **Number Formatting:** Financial values consistently use the Indian Lakh/Crore system (e.g., ₹ 1.50 Cr, ₹ 50.00 Lakh).
*   **Risk-Free Rate (Rf):** Based on `{clean_source(rf_src)}`. Dynamic fetch attempted using `{rf_proxy_ticker}`.
*   **Expected Market Return (Rm):** Based on `{clean_source(rm_src)}`.
*   **Market Risk Premium (MRP):** Derived (`Rm` - `Rf`).
*   **Beta (β):** Based on `{clean_source(beta_src)}`.
*   **Cost of Debt (Kd):** Calculated using `{kd_method}` method. Result bounded between {format_percent(MIN_COST_OF_DEBT)} and {format_percent(MAX_COST_OF_DEBT)}.
*   **Credit Spread:** Based on `{clean_source(cs_src)}` (used if Kd is calculated via Rf + Spread method).
*   **Tax Rate:** Effective tax rate based on `{tax_src}`. Result bounded between {format_percent(MIN_TAX_RATE)} and {format_percent(MAX_TAX_RATE)}.
*   **FCFF Calculation Method:**
    *   4-year average FCFF was pre-calculated for methods 0-4.
    *   User selected the method based on these averages (`Selected: {clean_source(fcff_method_src)}`).
    *   The 4-year average of the *selected method* (`{basis_meth}`) serves as the basis (T=0 FCFF) for projection.
    *   Historical FCFF values used for CAGR calculation were re-calculated using *only* the selected method.
    *   **Projection halts if the calculated 4-year average basis for the selected method is not strictly positive.**
*   **Growth Rates:**
    *   **Initial Growth:** Calculated as **`{init_g_fmt}`**.
        *   Source: `{growth_src}`.
        *   Method: Geometric Mean of 3-Year Revenue CAGR and 3-Year FCFF CAGR (calculated using the selected FCFF method's historical results). **Stock Price CAGR is NOT used.**
        *   Fallback: If GM fails or no valid CAGRs, defaults to `{def_init_g_fmt}`.
        *   Bounds: Final rate capped between `{min_init_g_fmt}` and `{max_init_g_fmt}`.
    *   **Terminal Growth (g):** Calculated as **`{term_g_fmt}`**.
        *   Method: Minimum of [Config Cap ({format_percent(TERMINAL_GROWTH_RATE_CAP)}), WACC * 0.60, Final Initial Growth ({init_g_fmt}), Rf * 0.90].
        *   Constraints: Floored at {format_percent(MIN_GROWTH)}, and ensured to be less than WACC ({format_percent(discount_rate)}).
*   **Projection Period:** Free Cash Flow to Firm (FCFF) projected for **{FORECAST_YEARS} years** beyond the basis year. Growth decays linearly from initial to terminal rate over this period.
*   **Capital Structure (for WACC Weights):** Weights based on Market Value of Equity (`Source: {clean_source(eq_src)}`) and Total Debt (`Source: {kd_method}`). Weights calculation method: `{clean_source(wgt_src)}`.
*   **Industry Benchmarks (for Ratios):** Comparison uses benchmarks from `{clean_source(bench_src)}`.

**Limitations & Considerations:**

*   **Model Sensitivity:** DCF valuations are highly sensitive to input assumptions, particularly WACC (driven by Rf, Beta, MRP, Kd), growth rates (initial and terminal), and the chosen **FCFF calculation method**. Small changes can significantly impact the result.
*   **FCFF Method Choice Impact:** The specific FCFF formula selected by the user directly influences the historical FCFF values, the resulting CAGR, the projection basis, and ultimately the valuation. Method 4 (EBITDA - Tax) is a simplified proxy and may differ substantially from others.
*   **Dependency on Positive Basis:** The model requires the chosen FCFF method to yield a positive 4-year average to proceed with valuation. If no method yields a positive average, the valuation halts.
*   **Input Quality & Fallbacks:** The reliability of the valuation depends heavily on the accuracy of user inputs (Rf, Rm, Beta, Spread, Benchmarks) and the appropriateness of default values or calculated fallbacks (e.g., derived tax rate, implied Kd, default growth).
*   **Historical Data Limitations:** Calculations based on historical data (CAGR, averages) are subject to the quality, availability, and length of the historical record provided by Yahoo Finance. Short or volatile histories reduce reliability.
*   **CAGR & Geometric Mean:** These methods have inherent limitations, especially with volatile or negative cash flows. The GM approach used here only considers Revenue and selected FCFF CAGR.
*   **Dynamic Data Fetching:** Relies on the `yfinance` library and Yahoo Finance API, which may experience downtime, data changes, or inaccuracies.
*   **Model Simplifications:** Assumes linear decay of growth, constant tax rates and capital structure during the projection (implicitly), and uses a single terminal growth rate in perpetuity. Ignores many qualitative factors (management, moat, etc.).
*   **Data Completeness:** Missing data points in financial statements can lead to calculations failing or relying on potentially inaccurate fallbacks (e.g., assuming zero depreciation or inventory).

**Disclaimer:**
This analysis is generated based on publicly available data and user-defined assumptions. It is intended for **educational and illustrative purposes only**. It does **NOT** constitute investment advice, solicitation, or recommendation to buy, sell, or hold any securities. Financial models, including DCF, have inherent limitations and are sensitive to assumptions. **Always conduct your own thorough due diligence** and consider multiple valuation methods before making any investment decisions. Consult with a qualified financial advisor for personalized advice. The creators of this script are **NOT liable** for any investment decisions made based on this output.
"""))











````
